**Tidal Evolution**



Since we know the Moon is drifting away, we can avoid the sign confusion and reduce the equation we need to this:

$$ \frac{da}{dt} = + \frac{3k_{2p}}{Q} \frac{m_s}{m_p} \frac{R_p^5}{a^{11/2}} \sqrt{G(m_p+m_s)} $$

The information we have is $\frac{da}{dt} = 0.0254 m/y$. Yes, we converted to meters. Never use imperial if youc an help it. We'd like this as a change in meters per second, though, so we convert to the much smaller 8.053e-10 m/s. 

The only value we don't really know here is k, which we assume is known as the notes give it as about 0.3. Moon mass is 7.348e22 kg, earth mass is 5.972e24 kg, earth radius is 6.371e6 m, moon semimajor axis is 3.844e8 m, and G is 6.674e-11. All together this becomes:

$$ (8.043e-10) = (1.419e-8)/Q $$
$$ \Rightarrow Q = 17.6$$

The order-of-magnitude estimate for Earth's Q is 10. This is close enough, as every value for Q given in the notes is always a direct power of ten. 

If we take that value for Q, we can create differential equation of the form y'=y, and integrate it to get total time. The equation itself is:

$$ \frac{da}{dt} = \frac{2.335e39}{Qa^{11/2}} = \frac{1.327e38}{a^{11/2}} $$

We probably don't need to solve the differnetial equation itself and can just go with a normal solution, but we felt the need to. And by that we man we had an online calculator do it for us. 

$$ a = \left( \frac{13(1.327e38)t+C}{2} \right) ^{2/13} = \left( \frac{(1.725e39)t+C}{2} \right)^{2/13} $$

Where C is the integration constant. It is related to the synchronous radius as we need the semimajor axis to equal that at t=0. 

Current a is 3.844e8 m. The synchronous radius is the radius of geosynchronous orbit at 3.580e7 m. 

First, we find C. At t=0 we need to solve $ (3.508e7) = (C/2)^{2/13} $ which gives a C of 2.208e49. 

So now we need to solve for t in 
$$ (3.844e8) = \left( \frac{(1.725e39)t+(2.308e49)}{2} \right)^{2/13} $$
$$ \Rightarrow (6.325e55) =  \frac{(1.725e39)t+(2.308e49)}{2} $$
$$ \Rightarrow (1.265e56) = (1.725e39)t $$
$$ \Rightarrow (7.334e16) = t $$

This is 2.325e9 years. 

It is expected that the moon formed about 4.5e9 years ago.
