# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$



The maximum principle states that **the maximum (and minimum) temperatures occur on the boundary**. And that they don't occur in the interior. This only occurs for steady-state for the Laplace equaiton. We are told that this leads directly to Laplace's equation $\nabla^2 u = 0$ with a specified boundary is well posed, but notably this is not proven to us. 

So first, we shall prove that, and then use it to show that Poisson's equation is also unique. 

There isn't much math in the proof, though, it's largely an argument, but we do have to note that the maximum principle comes from the mean value theorem, which states that the temperature at a point equals the average temperature from a circle around it. This means that the boundary determines the temperature of the point in the center of the object, so if the boundary is defined, we know what that temperature is. Thus every temperature is rigidly determined by a *disc* that expands around it until it reaches the boundary. Any symmetry that exists in the problem is indistinguishible, so "identical rotations" do not count as other solutions. Therefore the Laplace equation is unique, as the boundary rigidly defines the interior. Granted this isn't particuarly practical if the boundary is something other than a circle, but the boundary is guaranteed to INCLUDE circles, and every point that isn't the center also has circles coming out of it, forcing everything to equal. 

Now, the question is how do we apply this to the Poisson equation? It's not zero everywhere. However, if we can show the Mean Value Theorem holds, then the maximum principle holds, and then the solution is unique. But DOES the Mean Value Theorem hold? The problem is we can't use separation of variables, since the function g can be freely defined, so we can't split things up into angle and radius. And seeing how the solution of the Poisson Equation isn't stated in the book until chapter 8, we assume that we are not expected to find that. 

So, how can we justify that the maximum principle applies to Poisson's equation as well? Well, one thing we do know that Laplace and Poisson share is that everything has to be continuous, therefore the more genreal form of the Mean Value Theorem states that when traveling from one point to another, the middle temperature *must* be reaches in the path. This works no matter what path is taken to get from point to point, so on a circle, going from any point on one end to the other end *must* hit the value between those two points. To make it simple, let's only consider paths that go through the center, since that's the point we're concerned with. 

Now, since the average tempreature between any two points on the circle must exist within the circle, the average temperature for the entire circle must also exist within the circle, as there has to be a highest and lowest temperature on the defined boundary, meaning we can draw a line between those two poitns that always must pass through the total average, no matter what. 

So we've proven the average tempreature must exist somewhere in the circle. It can exist in multiple places, naturally, but now we seek to show that the center of the disc must have the average. Let us think of the disc as a series of concentric circles. Each and every last one of them must have an average temeperature value in them somewhere. We can start with the center itself, a single point, it has no boundary and thus is its own average temperature. If we apply a single infinitesimal layer to this point, drawing a "line" through the layer must go through the center, so it has to contain the average of all the points around it. Which means that the average temperature does not change from adding a single layer. So if we add another layer, the average temperature does not change, and so on and so forth. It does not matter that there are now more ponts between the center and the edge--we can always draw a line that goes through the center, and it has to cross through all the smaller circles to get there. Adding more circles cannot change the average temperature so long as everything is continuous.

This we have shown that the Mean Value Theorem, as it is described aplying to the Laplace Equation, actually applies to any two dimensional situation that is continuous. Now we can have the actual values of tempreature fluctuate quite a bit, but if we have to keep continuity, the center must be the average. This also makes physical sense, since in the case of heat or tempreature or anything really it has to "flow" from the boundary to the center, which feels the influence of all portions of the boundary equally. 

Anyway, since the Mean Value Theorem applies, the maximum principle applies, and the solution must be unique. 

NOTE: we were tempted to just take what the problem said at face value and assume the maximum principle applied, but we weren't comfortable with this, so we went a step further back. 

Secondary note: One can easily think of physical situations, though, where the center does not contain the average value... except this is, by nature, not an *eqilibrium* state. If you have a source in the center of a circle with cooler stuff around it, it's going to heat the rest of it up over time!

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$



Assuming "the solution is the same" led to annoyance, so we're going to solve it all the way through to identify *what* exactly that sign there changes. It shouldn't be much of anything, but the result originally had it be well-posed at the end, so we're going through it all the way.

First of all, there's not too much different in the separated equations.

$$ \frac{d^2 X}{dx^2} +X \lambda = 0$$
$$ \frac{dT}{dt} - T \kappa \lambda =0$$

The difference there is the sign flip in the temporal equation. Note that there is absolutely NO change to the physical solution. Thus the result for it is sitll the same.

$$ X(x) = C sin \frac{n\pi x}{L} $$

Now as for the *temporal* equation, the solution is evident just from looking at it. 

$$ T(t) = C e^{\kappa\lambda x} $$

Now due to the physical solution being the same we know what $\lambda$ is and so...

$$ T(t) = C e^{\kappa(\frac{n\pi}{L})^2x} $$

Pretty standard. Via superposition we assemble the total

$$ u(x,t) = \sum_{n=1}^\infty B_n sin\frac{n\pi x}{L} e^{\kappa (n\pi/L)^2 t} $$

The only effective change is the flipping of the sign in the exponent. Notably this does lead to immediate issues of *things exploding* so we can already tell this is going to be an issue. One might argue that we can get around this exploding by changing the sign of $\lambda$, but we actually can't do that! The physical part of the differential equation prodces cosh and sinh functions when solved for negative $\lambda$, and as we know sinh and cosh cannot meet the boundary conditions we have perscribed. Thus, our exponential is forced to be positive. And will quite easily explode to infinity as t gets larger. However, we do need to show that the rest of the function produces a difference that this exploding can magnify

We note there is no f(x) in the solution. Well, that's because it's hiding inside the coefficients.

$$ B_n = \frac 2L \int^L_0 f(x) sin\frac{n\pi x}{L} dx $$

If we make the adjustment...

$$ B_n' = \frac 2L \int^L_0 f(x) sin\frac{n\pi x}{L} dx + \frac 2L \int^L_0 \frac 1m sin\frac{m\pi x}{L} sin\frac{n\pi x}{L} dx $$

Now the only term we care about is the last one, for that indicates the "change" brought on by the 1/m situation. Due to orthogonality only m=n provides anything extra.

$$ \frac {2}{Lm} \int^L_0 sin^2\frac{m\pi x}{L} dx = \frac {2}{Lm} \frac{L}{2} = \frac 1m $$

Which is a small difference because m is specificed to be large. However, even a tiny difference will explode exponentially as t gets larger. It doesn't matter how small 1/m is, if you wait enough time, the solution erupts catastrophically. The fact that there is ANY difference at all in the initial solution means, with time, everything goes wrong. Now if we didn't need to solve this problem for all time, we would be fine, but the fact of the matter is that we do. Exponential growth is a killer.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$



The separation of variables is the same as always.

$$ \frac{d^2 \phi}{dx^2} +\phi \lambda = 0$$
$$ \frac{dG}{dt} + G \kappa \lambda =0$$

Today, though, we notice something odd, very odd indeed, in the boundary conditions. Not only do we have a case where the arbitrary condition is on position instead of time, but the other condition is quite strange! 

The rewrites of the conditions are:

$$ \phi(x) = f(x) $$
$$ G(0) = 1 $$
$$ \phi(0) = 0 $$
$$ \phi(1) + \phi'(1) = 0 $$

We said the last one was kind of odd. 

Regardless, we can get the general solutions. However we no longer know $\lambda$ so we'll list ALL possible solutions. For time dependence, the sign of $\lambda$ doesn't fundamentally change it's structure.

$$ G = Ce^{-\kappa\lambda t} $$

The boundary conditions we have assigned force C=1. Though this is technically arbitrary, it just makes it ieaseir if f(x) is $\phi(x)$. Now, the position function will have different forms:

Positive $\lambda$:

$$ \phi = Asin(\sqrt{\lambda}x) + Bcos(\sqrt{\lambda}x) $$

Zero $\lambda$:

$$\phi = A + Bx$$

Negative $\lambda$:

$$\phi = Acosh(\sqrt{-\lambda} x) + Bsinh(\sqrt{-\lambda} x)$$

Now to go further we need to figure out $\lambda$. But that seems to be the work of part b), so we shall stay there. 

We can remove the zero case rather quickly. We need somedthing that is zero at the x=0 situation, so we already remove the A constant and have $\phi = Bx$. Well Bx cannot equal -B, so the derivative self-cancel fails, so the zero case is impossible.

Now, why is the hyperbolic trig case impossible? It's simple, sinh and cosh are each other's derivatives. There is no way to produce a "canceling" by taking derivatives, so the negative result is impossible. Also on the positive side of the domain, which is where x=1 has to lie, they are both positive so there's not even a slight chance. Definitely not possible.

Thus, all that remains is the positive case. Just to keep our sanity, let's check to see if it works. 

$$ \phi = Asin(\sqrt{\lambda}x) + Bcos(\sqrt{\lambda}x) $$

First of all we have to have x=0 make this equal zero, so the cosine goes away. 

$$ \phi = Asin(\sqrt{\lambda}x)$$

Now, this derivative is cos. sin and cos *do* regularly have points where they cancel each other out, sin + cos has plenty of zeroes. Usually at this point we would try to find out what those zeroes were so we could find $\lambda$ explicitly but we have been informed that is not possible, so we'll just be happy knowing that the form of our solution is:

$$ u(x,t) = \sum_n A_n sin(\sqrt{\lambda_n}x) e^{-\kappa \lambda_n t} $$

Anyway we now need to declare what the $\lambda$ values satisfy. 

$$ sin(\sqrt{\lambda} x) + \sqrt{\lambda}cos(\sqrt{\lambda}x) = 0 $$
$$ \Rightarrow sin(\sqrt{\lambda}) + \sqrt{\lambda}cos(\sqrt{\lambda}) = 0 $$

The x variable is supressed since it's evaluated at x=1 for this condition. The constants A have been removed since they aren't necessary. We know solutions exist since sinx+cosx have many zero solutions, and the adjustment of $\sqrt{\lambda}$ just expands or contracts the wavefunction, moving those zeroes back and forth. 

But, as mentioned, we can't solve for them. So yeah. 

We already wrote it in the correct form in the above part. We used geogebra to get soem quick approximations to see if we were in the right ballpark, solving for $\sqrt{\lambda} \approx 2, 5, 8, 11$ These are not precise but they give us good ballparks, and point out that there's probably a pattern. They also provide nice guesses to feed the program! ...We realize that we could have done this from the matlab plot, but we didn't at the time. 

Regardless, coding this up will not be difficult.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$



The sine series assumes ODD projection across the y axis. Which is is to say, we end up with f(x) = -1-x on [-1, 0)

Anyway now that we actually have a full description, we can find the fourier sine series. 

$$ f(x) \text{~} \sum_{n=1}^\infty B_n sin\frac{n\pi x}{L} $$
$$ B_n = \frac 2L \int^L_0 f(x) sin\frac{n\pi x}{L} $$

Note that we actually only integrate from 0 to L. This is due to symmetry of the sine function and the symmetry we "forced" around the x-axis. If we integrated in full, we would have:

$$ B_n = \frac 1L \int^L_{-L} f(x) sin\frac{n\pi x}{L} $$

But symmetry lets us ignore this. 

Anyway, actually taking the integral...

$$ B_n = 2 \int^1_0 (1-x) sin(n\pi x) $$
$$ = 2 \left( [-\frac{1}{n \pi}cos\frac(n\pi x)]_0^1 - [\frac{1}{n^2\pi^2}sin\frac(n \pi x) - x\frac{1}{n \pi}cos(n\pi x))]_0^1 \right) $$

$$ = 2 \left( -\frac{1}{n \pi}cos(n\pi) + \frac{1}{n\pi} - 0 + \frac{1}{n \pi}cos(n\pi) \right) $$
$$ = \frac{2}{n\pi} $$

All the coefficients nicely cancel, it seems, giving us this simple term. Which means our answer is...

$$ f(x) \text{ ~ } \sum_{n=1}^\infty \frac{2}{n\pi} sin(n \pi x) $$

Geogebra confirms this works. 



This is adapted from the same code we made back in homework 3. Which is why the total function is V. 

```
x=-3:0.001:3;
N=20;
V=0;

for n=1:N
    V=V+2/(n*pi)*sin(n*pi*x);
end
    
plot(x,V);
title('Fourier series for f(x)');
xlabel('x'); 
ylabel('~f(x)');
hold
```



If we run it with N=200...



Where we can clearly see the spikes at the discontinuities. 

Clearly every individual term goes to 0 when x=0, so the series itself has to also go to x=0. This is standard for a sine series, we can't think of a situation where it could do anything else, for the sine terms always go to 0 at x=0.

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$



The cosine series assumes EVEN projection across the y axis. This means on the other side we have x+1. But like with the sine series, we have a shortcut rather than doing the entire thing.

$$ f(x) \text{ ~ } \sum_{n=0}^\infty A_n cos\frac{n\pi x}{L} $$
$$ A_0 = \frac 1L \int_0^L f(x) dx$$
$$ A_n = \frac 2L \int_0^L f(x) cos\frac{n\pi x}{L} dx $$

So in our case our first coefficient...

$$ A_0 = \int_0^1 (1-x) dx = 1 - \frac12 = \frac 12 $$

Simple. As for the other part...

$$ A_n = 2 \int_0^1 (1-x) cos(n\pi x) dx$$
$$ = 2 \left( [\frac{1}{n\pi}sin(n\pi x)]_0^1 - [\frac{1}{n^2\pi^2}cos(n\pi x)+x\frac{1}{n\pi}sin(n\pi x)]_0^1\right) $$
$$ = 2 \left( 0 + \frac{1}{n^2\pi^2}cos(n\pi) + 0 + \frac{1}{n^2\pi^2} \right) $$
$$ = \frac{2}{n^2\pi^2}(1-cos(n\pi))$$

Practically, this coefficient flops from the parenthetical term being 0 and 2. 

$$ f(x) \text{ ~ } \frac12 + \sum_{n=1}^\infty \frac{2}{n^2\pi^2}(1-cos(n\pi)) cos(n\pi x)  $$



Shockingly precise. 

Second verse, same as the first.

```
x=-3:0.001:3;
N=20;
V=0.5;

for n=1:N
    V=V+2/(n*n*pi*pi)*(1-cos(n*pi))*cos(n*pi*x);
end
    
plot(x,V);
title('Fourier series for f(x)');
xlabel('x'); 
ylabel('~f(x)');
hold
```



Considering how there was harldy any noticalbe features on the low-order GeoGebra plot, we are not at all surprised to see what we are going to call "scarily precise" results. At low values there aren't even any weird oscillations, just the rigid spikes, which is the same as saying no Gibbs Phenomenon. 

N=200 has no visible difference. 



See, from the picture it's clear it converges to 1. But how, though? Well, let's put in x=0 and examine the series we get.

$$ f(0) \text{ ~ } \frac12 + \sum_{n=1}^\infty \frac{2}{n^2\pi^2}(1-cos(n\pi))  $$

If we restrict n to odd only...

$$ f(0) \text{ ~ } \frac12 + \sum_{n=1,3,5,...}^\infty \frac{4}{n^2\pi^2} $$

Pull out the constants...

$$ f(0) \text{ ~ } \frac12 + \frac{4}{\pi^2}\sum_{n=1,3,5,...}^\infty \frac{1}{n^2} $$

Now, just to make sure we weren't crazy, we ran that infinite series through Wolfram. It does, in fact, converge to $\frac{\pi^2}{8}$, making the entire series converge to 1. But we consider this merely a check, not a solution, WHY does the series converge to that? 

It is somewhat well known that the series $\frac {1}{n^2}$ goes to $\frac{\pi^2}{6}$, but that includes the even squares as well, not just the odd ones. Looking into how this was proven proved to be a minefield, so we seek another way to justify it. 

Now, what we can do is rather easily find what the EVEN sum goes to, since we can represent it as $\frac{1}{4n^2}$] meaning it converges to $\frac{\pi^2}{24}$ since the 1/4 can be pulled out. 

The total sum minus the even sum would equal the odd sum. $\frac{\pi^2}{6} - \frac{\pi^2}{24} = \frac{4\pi^2}{24} - \frac{\pi^2}{24} = \frac{3\pi^2}{24} = \frac{\pi^2}{8}$ Which is exactly what we were looking to prove. Thus, the overall value of the series converges to 1.

# Problem 6 \[Back to [top](#toc)\]
$$\label{P6}$$



So the full series is given by:

$$ f(x) \text{ ~ } A_0 + \sum_{n=1}^\infty A_n cos\frac{n\pi x}{L} + \sum_{n=1}^\infty B_n sin\frac{n\pi x}{L} $$
$$ A_0 = \frac{1}{2L} \int_{-L}^{L} f(x) dx $$
$$ A_n = \frac{1}{L} \int_{-L}^{L} f(x) cos\frac{n\pi x}{L} dx $$
$$ B_n = \frac{1}{L} \int_{-L}^{L} f(x) sin\frac{n\pi x}{L} dx $$

Rather than writing it all out in full we'll just breeze over it, relying on looking them up rather than just doing it every time since these integrals are essentially giong to be the same as the previous two problems. In fact, it's going to be basically identical for the most part, as all the integrals from -1 to 0 are zero since the function IS zero there.

$$ A_0 = \frac{1}{2} \int_{0}^{1} (1-x^2) dx $$
$$ A_n = \int_{0}^{1} (1-x^2) cos(n\pi x) dx $$
$$ B_n = \int_{0}^{1} (1-x^2) sin(n\pi x) dx $$

And skipping to the end here...

$$ A_0 = \frac{1}{3} $$
$$ A_n = - \frac{2}{n^2\pi^2} cos(n\pi)  $$
$$ B_n = \frac{- 2}{n^3\pi^3} cos(n\pi) + \frac{n^2\pi^2 + 2}{n^3\pi^3} $$

Rather than writing everything out, we just assume the reader knows how to substitute the proper things in. 

$$ f(x) \text{ ~ } \frac13 + \sum_{n=1}^\infty A_n cos(n \pi x) + \sum_{n=1}^\infty B_n sin(n \pi x) $$

And in GeoGebra we get:



Third verse same as the first! Though in this one A and B are declared separately for neatness sake. 

```
x=-3:0.001:3;
N=20;
V=1/3;
A=0;
B=0;

for n=1:N
    A = -2/(n*n*pi*pi)*cos(n*pi);
    B = -2/(n*n*n*pi*pi*pi)*cos(n*pi) + (n*n*pi*pi + 2)/(n*n*n*pi*pi*pi);
    V=V+A*cos(n*pi*x)+B*sin(n*pi*x);
end
    
plot(x,V);
title('Fourier series for f(x)');
xlabel('x'); 
ylabel('~f(x)');
hold
```

And this clearly reveals our issue with not wanting to deal with powers. Probably should have just n^2 but... it's there now and it works. 




Clear Gibbs Phenomena in N=200. 



Note that when x=0 the sine term goes completley away. This leaves us with

$$ = \frac13 + \sum_{n=1,3,5,7...}^\infty \frac{2}{n^2\pi^2} - \sum_{n=2,4,6,8...}^\infty \frac{2}{n^2\pi^2} $$

We already know what the even and odd series are from the previous problem.
