# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

First of all, we write this out in simpler form.

$$ u_t = ku_{xx} + k $$

Since it is stated we should assume that Q = k. We do not know what the boundary conditions are in specific aside from the fact that they are constant. They could be the same, they could be different, they could be zero, we can assume nothing about the temperature at the ends aside from the fact that it is fixed.

We seek an equilibrium solution, that is, one that satisfies no time dependence. 

$$ 0 = ku_{xx} + k \rightarrow 0 = u_{xx} + 1 \rightarrow u_{xx} = -1 $$

As well as the A adn B boundary conditions. First, the general solution for this sort of equation. In two derivative steps the result has to be a constant, so obviously we have an $x^2$ type function with possible lower terms in front.

$$ C_1 + C_2 x + C_3 x^2 $$

The second derivative of this  is $2C_3$, which means $C_3$ better equal -1/2. 

$$ C_1 + C_2x - x^2/2 $$

Now we need to match boundary conditions. at x=0, we need to equal A. Easy enough, set $C_1$ to A. 

$$ A + C_2x - x^2/2 $$

How at x=L, the entire thing needs to equal B. So let's solve for $C_2$ with that requirement.

$$ A + C_2L - L^2/2 = B$$
$$ \Rightarrow C_2L = B - A + L^2/2 $$
$$ \Rightarrow C_2 = \frac{B - A + L^2/2}{L} $$

And so our equilibrium solution is...

$$ A + \frac{B - A + L^2/2}{L}x - x^2/2 $$

Which has a second derivative of -1, equals A at x=0, and equals B at x=L. Which is all the criteria we need for the equilibrium solution. Thus, the equilibrium solution exists. 

Now let us say our solution u is the total solution, and let v be a function that is left after the equlibrium solution is subtracted. That is, the "time-based corrective."  $v = u - u_E$. We note that v does *not* solve the original differential equation, what it solves instead is:

$$ v_t = kv_{xx} $$

Since the "+k+ term is taken care of by the equilibrium portion. Furthermore, since u always equals $u_E$ at the boundaries, their subtraction boundary condition is zero. Thus, v itself becomes a differential equation we are somewhat familiar with: the heat equation without a source term that is zero at the ends. This is an extremely common problem, the result is well known.

$$ v = \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L}e^{-k(n\pi/L)^2t}$$

Since we are given a very general initial condition, the coefficietns are also easy to find based on past work.

$$ f(x) - u_E(x) = v(x,0) $$
$$ \Rightarrow A_n = \frac{2}{L} \int_0^L [f(x) - u_E(x)]sin\frac{n\pi x}{L} dx$$

So if we write out our answer in full...

$$ u(x,t) = v(x,t) + u_E(x) $$
$$ v(x,t) = \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L}e^{-k(n\pi/L)^2t}$$
$$ A_n =  \frac{2}{L} \int_0^L [f(x) - u_E(x)]sin\frac{n\pi x}{L} dx $$
$$ u_E(x) = A + \frac{B - A + L^2/2}{L}x - x^2/2 $$

The last question is what happens as t approahces infinity. Well, the time-dependent v(x,t) gets smaller and smaller until it approaches zero for each of its terms (so long as k is positive which we think we can assume at this point). This means v as a whole goes to zero, and thus:

$$ u = u_E $$

Which is to say at large t u approaches equilibrium. Which is the case for *virtually all* heat problems, and is quite expected.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

First of all, let's simplify the appearance a bit.

$$ u_t = u_{xx} + e^{-2t} sin5x $$

Clearly nonhomogenous and our extra term is time dependent, so this will be a bit more difficult than the previous one. At least the boundary conditions at the ends are constant, that will help us out considerably. The initial condition that sets everything to zero is simplifying as well. 

Fortunately, we can find an equilibrium solution here. Knowing what we do about the behavior of these sorts of equations, we can in fact find a section where the time dependence is irrelevant. Even though our right-most term depends on t, at large t it might as well not be there at all, and so we can find an equilibrium there. 

$$ 0 = u_{Exx} $$

The general form for such a thing subject to constant boundary conditions is:

$$ u_E = A + \frac{B-A}{L}x $$

In our case, A=1, B=0. 

$$ u_E = 1 - \frac{x}{L} $$

This is the equilibrium solution, where the equilibrium exists as "large t." Since $u_E$ is linear in x, that means our function $v = u - u_E$ has to match the same differential equation u does, as $u_E$ provides absolutly nothing to the differential equation. What it does do is make it so the boundary conditions on v are both zero.

$$ v_t = v_{xx} + e^{-2t}sin5x $$ 

Where our boundary conditions are v(0,t)=0, and v($\pi$,t)=0. The form of the right-most term gives us a hint as to the form of the function v: an exponential times a sine function of some kind, preferably one that can add to the old sine function once its derivative is taken. So let's see what the derivatives of that right-most term are. With respect to t...

$$ -2e^{-2t} sin5x $$

And with respect to x twice.

$$ - 25 e^{-2t} sin5x  $$

So essentially we need to solve for A in...

$$ -2A = -25 A + 1 $$
$$ \Rightarrow 23 A = 1 $$
$$ \Rightarrow A = 1/23 $$

Which seems like an odd number, but when we check, it does in fact solve the equation.

$$ -\frac{1}{23} 2 e^{-2t}sin5x = - \frac{1}{23} 25 e^{-2t}sin5x + e^{-2t}sin5x $$ 
$$ -\Rightarrow \frac{2}{23} = - \frac{25}{23} + \frac{23}{23} $$
$$ -\Rightarrow \frac{2}{23} = -\frac{2}{23} $$

So we have *a* solution for v. 

$$ v = \frac{1}{23} e^{-2t} sin5x $$

As one can recall from previous differential equations though, just because we found one solution does not mean we found them all. We have the particular solution, now we need the general homogenous solution.

$$ v_t = v_{xx} $$

Which has a well known solution.

$$ \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L}e^{-(n\pi/L)^2t}$$

And so the total equation for v is:

$$ v(x,t) = \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L}e^{-(n\pi/L)^2t} + \frac{1}{23} e^{-2t} sin5x $$

And the actual entire solution for u is...

$$ u(x,t) = \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L}e^{-(n\pi/L)^2t} + \frac{1}{23} e^{-2t} sin5x + 1 - \frac{x}{L} $$

Now, we actually happen to *have* initial conditions! So we can set t to be 0 and get:

$$ 0 = \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L} + \frac{1}{23} sin5x + 1 - \frac{x}{L} $$
$$ \Rightarrow - \frac{1}{23} sin5x - 1 + \frac{x}{L} = \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L} $$
$$ \Rightarrow A_n = \frac{2}{L} \int_0^L [- \frac{1}{23} sin5x - 1 + \frac{x}{L}] sin\frac{n \pi x}{L}dx $$

Now this can actually be evaluated. First of all, we do have a value for L, it is $\pi$.

$$ \Rightarrow A_n = \frac{2}{\pi} \int_0^\pi [- \frac{1}{23} sin5x - 1 + \frac{x}{\pi}] sin(n x)dx $$

Well that's interesting! We have an orthogonality here, the sine funcitons will cancel each other at all n values except n=5. Interesting. Let's leave that integral for last, then, and just deal with the other terms first. Sucn integrals are pretty trivial. The constant term vanishes due to the bounds being $\pi$. The term with x in it does not

$$ \int_0^\pi \frac{x}{\pi} sin(nx) dx = \frac{1}{\pi} \left[ \frac{1}{n^2} sin(nx) - \frac{1}{n} x cos(nx)  \right]_0^\pi = - \frac{\pi}{n} cos(n\pi) $$

the cosine term will flip from +1 to -1 depending on n. Now, as for the orthogonality, it vanishes for every term *except* 5. 

$$ \int_0^\pi - \frac{1}{23} sin^2(5x) dx = -\frac{1}{23} \frac{\pi}{2} $$

Thus...

$$ \Rightarrow A_n = - \frac{2}{\pi} \frac{\pi}{n} cos(n\pi) = - \frac{2}{n} (-1)^n $$

EXCEPT for n=5, where

$$ \Rightarrow A_5 = - \frac{2}{\pi} (\frac{1}{23} \frac{\pi}{2} - \frac{\pi}{5}) = - \frac{2}{23} + \frac{2}{5} = \frac{36}{115} $$

And so if we are to write out our full solution:

$$ u(x,t) = \sum_{n=1}^\infty A_n sin\frac{n\pi x}{L}e^{-(n\pi/L)^2t} + \frac{1}{23} e^{-2t} sin5x + 1 - \frac{x}{L} $$
$$ A_5 = \frac{36}{115} $$
$$ A_n(n \neq 5) = - \frac{2}{n} (-1)^n $$

Which is highly unusual in form, but then again we did have a highly unusual setup for the problem. But now our boundary conditions have been met, our initial conditions have been met, and our coefficients have been determined. We have the solution.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

In general, fourier transforms of derivatives produce terms of $i\omega$. The extra term at the end is decidedly odd, but let's roll with it.

Green's formula is given in 5.5.8

$$ \int_a^b [uL(v) - vL(u)]dx = p(uv_x - vu_x)_a^b $$

Note that the p comes from $L = (py_x)_x + qy$. The exact adjustment here depends on what we declare L to be. 

Let's declare L to be the second derivative, thus p=1 and q=0 (not that the q part matters). Now, we originally got this solution by working backward, so when we work forward it looks like we're doing a lot of things for no reason. Start with the Fourier transform of the second derivative. Let it equal T. 

$$ \frac{1}{2\pi} \int_{-\infty}^\infty f_{xx} e^{i\omega x} dx = T $$

Now, for "apparently no reason," add the fourier transform of f alone to both sides times $\omega^2$.

$$ \Rightarrow \frac{1}{2\pi} \int_{-\infty}^\infty f_{xx} e^{i\omega x} dx + \omega^2F = T + \omega^2F $$

Since F is just an integral over the same bounds as the other one, and with the same leading coefficient to boot, we can just add their interiors.

$$ \Rightarrow \frac{1}{2\pi} \int_{-\infty}^\infty f_{xx} e^{i\omega x} + \omega^2 e^{i\omega x} fdx  = T + \omega^2F $$

Now if L is the second derivative, we have a second derifative acting on f and effectively a second derivative acting on the exponential. Therefore...

$$ \Rightarrow \frac{1}{2\pi}\left( e^{i\omega x} f_x - (e^{i\omega x})_xf\right)|_{-\infty}^\infty  = T + \omega^2F $$
$$ \Rightarrow \frac{e^{i\omega x}}{2\pi}\left( f_x - i\omega f\right)|_{-\infty}^\infty  = T + \omega^2F $$
$$ \Rightarrow \frac{e^{i\omega x}}{2\pi}\left( f_x - i\omega f\right)|_{-\infty}^\infty - \omega^2F = T $$

And since T is our fourier transform of the second derivative, we have now shown exactly what we sought to. 

Of note, so long as the function f goes to zero as x goes to infinity, we know that the evaluation has to go to zero, since the normal result of $\mathcal F f_{xx} = -\omega^2 F$

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$

The shift theorem is an easy one to state: the inverse fourier transform of $e^{i\omega \beta} F$ is $f(x-\beta)$

The convolution theorem is stated as follows

$$ h(x) = \frac{1}{2\pi} \int_{-\infty}^{\infty} f(x') g(x-x') dx' $$

Now what h is... we'll get there, but we'll need to keep the convolution in mind for later. However, before we get to the point we need to use it, we need to back up and consider the problem on infinite bounds.

So we should use Fourier methods rather than separation of variables--even though this can be separated, the spatial portion is a bit of a mess (albeit actually solvable). So, to start, we define our Fourier transform to be in the spatial variable and change the differnetial equation.

$$ \mathcal{F} [u_t] = k\mathcal{F} [u_{xx}] + c \mathcal{F} [u_x] $$

The x derivatives can be worked out somewhat easily using known rules. 

$$ \mathcal{F}[u]_t = - \omega^2 k\mathcal{F} [u] - i \omega c \mathcal{F} [u] $$

Treating the Fourier transform as the function we are solving for, we end up with a function of the form:

$$ U_t = (-\omega^2 k - i\omega c)U  $$

Which has a solution

$$ U = \mathcal F [u] = C(\omega) e^{(-\omega^2k - ic\omega)t} $$

Now, what is C? It's a leading coefficient "function" that is the initial condition of the transform, which is to say the transform of u(x,0). Which we can represent by just $F(\omega)$. 

$$ U = F(\omega) e^{(-\omega^2k - ic\omega)t} $$

Now in theory we could work with this, or we can go to the convolution theorem. We can already see potential use for the shift theorem, as we are "shifted" from the normal gaussian here. Neat.

To actually use the convolution, we need to identify both an F and a G. Fortunatly the F is already provided for us, and its inverse fourier transform is the initial conditions, as previously discussed: f(x). G in this case is our shifted Gaussian. Now, the inverse Fourier transform of a non-shifted Gaussian is:

$$ e^{-\beta \omega^2} \rightarrow \sqrt{\frac{\pi}{\beta}} e^{-x^2/4\beta} $$

Now, this is being shifted by a factor of $e^{-ic\omega t}$ in our case. The shift theorem tells us that this means "x" in the resultant will need to be replaced--the form given is $x-\beta$ for $i\omega \beta$, so in this case we replace x with $x + ct$. Naturally there is still a time-dependence, shocker. In conclusion:

$$ e^{(-\omega^2k - ic\omega)t} \rightarrow \sqrt{\frac{\pi}{kt}} e^{-(x+ct)^2/4kt} $$

So we are now armed to write out the form of the answer for actual u using the convolution theorem. 

$$ u(x,t) = \frac{1}{2\pi} \int_{-\infty}^\infty f(x') \sqrt{\frac{\pi}{kt}} e^{-(x-x'+ct)^2/4kt} dx $$

Now as we don't have an initial condition specified, this is as far as we go. 

Now a dirac delta function is very convenient! It pulls out the value of the function at x'=0 when integrated over all space. Thus, for that initial condition...

$$ u(x,t) = \frac{1}{2\pi} \sqrt{\frac{\pi}{kt}} e^{-(x+ct)^2/4kt} $$

The result is just the Gaussian. 

In order to get a good picture, we will plot at various values of t, with k=1 and c a variable we can adjust at will. 



Here we have a negative c, which allows the wave to move to the right in time. When c is positive, it moves to the left, and when c is 0 it doesn't go anywhere and just dissipates over time. 

Pretty clearly c represents the velocity, or is closely related to it. Larger mangitude c results in "faster" moving waves. Which is probably why it's called c since we'd bet in ideal physical conditions we'd get the speed of light out of it.
