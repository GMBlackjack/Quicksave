# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

We originally solved this by working with the known solution for u directly, but the second part of the problem implies there is another way. (Also we realized later that without boundary conditions that solution does not necessarily hold.) First of all, let's find x and t in terms of these new variables.

$$ x = \frac{\eta + \xi}{2} $$
$$ t = \frac{\xi - \eta}{2c} $$

Which we can then substitute into the differnetial equation.

$$ \frac{d^2 u}{d (\frac{\xi - \eta}{2c})^2} - c^2 \frac{d^2 u}{d(\frac{\eta + \xi}{2})^2} = 0$$
$$ \Rightarrow 4c^2 \frac{d^2 u}{d (\xi - \eta)^2} - 4c^2 \frac{d^2 u}{d(\eta + \xi)^2} = 0$$
$$ \Rightarrow \frac{d^2 u}{d (\xi - \eta)^2} - \frac{d^2 u}{d(\eta + \xi)^2} = 0$$
$$ \Rightarrow \frac{d^2 u}{d\xi^2 - 2d\eta d\xi +d\eta^2} - \frac{d^2 u}{d\xi^2 + 2d\eta d\xi +d\eta^2} = 0$$

Now this is the part where things get somewhat quesitonable. The only difference between the terms is the $\pm$ in the second part of the differential polynomial. The only way the subtraction can make the equality equal zero, however, is if this sign *doesn't matter*, that is both of the terms are absolutely equal. 

$$ \Rightarrow \frac{d^2 u}{d\xi^2 - 2d\eta d\xi +d\eta^2} = \frac{d^2 u}{d\xi^2 + 2d\eta d\xi +d\eta^2} $$

This can *only* be true of the $d\eta d\xi$ term is irrelevant, that is, that the derivative with respect to one variable and then aother is zero. Which is to say...

$$ \frac{d^2u(\xi,\eta)}{d\xi d\eta} = 0 $$

Which is exactly what we sought to show. When this is true there is no part of the function u that depends on $\xi$ and $\eta$ at the same time, and so our original relation becomes:

$$ \Rightarrow \frac{d^2 u}{d\xi^2 +d\eta^2} - \frac{d^2 u}{d\xi^2 +d\eta^2} = 0$$

Which is clearly true. 

The second part follows rather cleanly from the first. We know the funciton u (or Y) has no terms that depend on both $\xi$ and $\eta$, but that the function itself depends on both. Therefore, the function must consist of a sum of various terms of one or the other. Which is to say, u(x,t) = F(x+ct) + G(x-ct) where F and G are unknown functions.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

The best way to find F and G is just to solve the problem and see what remains. One thing we can say is that the genreal solution to the wave equation still applies:

$$ u(x,t) = \sum_{n=1}^\infty \left( A_n sin\frac{n\pi x}{L}cos\frac{n\pi ct}{L} + B_n sin\frac{n\pi x}{L} sin\frac{n \pi ct}{L}\right) $$

Though now it becomes important to know what the constant coefficients are, unlike the previous problem.

$$ A_n = \frac 2L \int_0^L f(x) sin\frac{n\pi x}{L} dx $$
$$ B_n \frac{n \pi c}{L} = \frac 2L \int_0^L g(x) sin\frac{n\pi x}{L} dx $$

Anyway, as per usual we can try to simplify this by applying the boundary conditions. First of all, in space, we need to have the funciton be 0 at t=0 and t=L. This is already met by both terms A and B, since they are sine functions, so no simplificaiton there. Furthermore, since we have both f(x) and g(x) terms A and B remain as well. So it looks as though we have a full set of everything. Annoying. However, we can split the entire thing up into the variables we want to use:

$$ u(x,t) = \sum_{n=1}^\infty \left( A_n \frac12 \left[ sin(\frac{n\pi}{L}(x+ct)) + sin(\frac{n\pi}{L}(x-ct)) \right] + B_n \frac12 \left[ cos(\frac{n\pi}{L}(x-ct)) - cos(\frac{n\pi}{L}(x-ct)) \right]\right) $$


Now, a cool thing is that we've done half of htis problem already. **Problem 3** in the previous homework had us solve for the situation where we only knew g(x) and f(x)=0. However, g(x) only influences the value of $B_n$, and eveyrything else remains the same. The solution for the $B_n$ terms was...



Which means for our situation we have:

$$ u(x,t) = \sum_{n=1}^\infty \left( A_n \frac12 \left[ sin(\frac{n\pi}{L}(x+ct)) + sin(\frac{n\pi}{L}(x-ct)) \right] \right) + \frac{1}{2c}\int_{x-ct}^{x+ct} G(\xi) d\xi $$

where G is the *odd extension of g(x)*. 

So now that we have G we can ignore that term and focus on finding F only. 

$$ \sum_{n=1}^\infty \left( A_n \frac12 \left[ sin(\frac{n\pi}{L}(x+ct)) + sin(\frac{n\pi}{L}(x-ct)) \right] \right) $$
$$ = \frac 12 \sum_{n=1}^\infty \left( \frac 2L \int_0^L f(x) sin\frac{n\pi x}{L} dx \left[ sin(\frac{n\pi}{L}(x+ct)) + sin(\frac{n\pi}{L}(x-ct)) \right] \right) $$
$$ = \frac 12 \sum_{n=1}^\infty \frac 2L \left(  \int_0^L f(x) sin\frac{n\pi x}{L} dx sin(\frac{n\pi}{L}(x+ct)) + \int_0^L f(x) sin\frac{n\pi x}{L} dx sin(\frac{n\pi}{L}(x-ct)) \right) $$

Note the definition of the odd extension of G. 

$$ G(x) = \sum_{n=1}^\infty \frac2L \int_0^L g(x') sin\frac{n \pi x'}{L} dx' sin\frac{n \pi x}{L} $$

Note that this is *exactly* what we have above, just for f(x) instead of g(x). Thus, if we let F be the odd extension of f...

$$ = \frac 12 \left( F(x+ct) + F(x-ct) \right) $$

Which, when we put it into our total relation:

$$ u(x,t) = \frac 12 \left( F(x+ct) + F(x-ct) \right) + \frac{1}{2c}\int_{x-ct}^{x+ct} G(\xi) d\xi $$

Is exactly what we sought. And, as we expected, along the way we determined what F and G were: the odd extensions of their respective functions f and g.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

So we are only a funciton of x and t, which means our separated versions are going to be X and T. Let u = XT.

$$ (XT)_{tt} + (XT)_{xxxx} + \delta (XT)_{t} + k(XT) = 0 $$
$$ \Rightarrow XT_{tt} + TX_{xxxx} + \delta XT_{t} + kXT = 0 $$
$$ \Rightarrow XT_{tt} + \delta XT_{t} + kXT = - TX_{xxxx} $$
$$ \Rightarrow T_{tt} + \delta T_{t} + kT = - T\frac1X X_{xxxx} $$
$$ \Rightarrow\frac1T T_{tt} + \delta\frac1T T_{t} + k = - \frac1X X_{xxxx} $$


The variables are seprated, the relation between them must be constant. Call it $-\lambda$ even though we really aren't sure of the sign that'll be most convenient yet. 

$$ T_{tt} + \delta T_t + k + \lambda T = 0 $$
$$ - X_{xxxx} + \lambda X = 0 $$

These are our separated differential equations. Let's take a moment to consider $\lambda$. If it's zero, well pretty clearly the result from X will not be able to satisfy the boundary conditions since there won't be an oscillatory result. It is not immediately obvious that negative $\lambda$ would cause a problem, however... but we'll address this later. For the moment, let's assume positive.

First, solve in generality without worrying about boundary conditions. The X version is solvable with the characteristic polynomial. 

$$ -r^4 + \lambda = 0 $$

Which basically amounts to $r^4 = \lambda$. Are there four ways to do this? Well, yes, actually. We have the positive fourth root, the negative fourth root, and those again just with imaginary numbers in the front. It's nic3e to be albe to just look at something and see waht it is. With the four roots, we can write out the solution. Let the fourth root of $\lambda$ be J so we don't get lost in notation.

$$ X = Ae^{Jx} + Be^{-Jx} + Ce^{iJx} + De^{-iJx} $$

And this is just a general solution. Now we should at this point convert the imgatinary terms into trig ones... 

$$ X = Ae^{Jx} + Be^{-Jx} + Ccos(Jx) + Dsin(Jx) $$

And we will not simplify it until later, when we apply boundary conditions. 

Now, for our other situation, we have the T equation. Unfortunately that k means it's non-homogeneous. Fortunately it is only second order. To do this we solve the homogeneous version first, which is to say:

$$ r^2 + \delta r + \lambda = 0 $$

Which will have two roots of some kind. The quadratic formula can give them to us in this case, $\frac{-\delta \pm \sqrt{\delta^2 - 4\lambda}}{2}$. We might be tempted to make use of $\delta$ being small, but we can only really say it is small with respect to k, and k is not here yet. (Once we solve for $\lambda$ we may be able to simplify.) So the homogeneous solution is:

$$ Ee^{\frac{-\delta + \sqrt{\delta^2 - 4\lambda}}{2}t} + Fe^{\frac{-\delta - \sqrt{\delta^2 - 4\lambda}}{2}t} $$

We note that, if $\delta$ is small compared to $\lambda$ this is very clearly giong to be imaginary and we'll probably end up converting to trig functions, but we have to wait for that. 

Now finding a particular solution to add to this is not an issue at all: we need to equal -k, and we have a $\lambda T$ term, so a simple $T = -\frac{k}{\lambda}$ will do it. And so...

$$ T = Ee^{\frac{-\delta + \sqrt{\delta^2 - 4\lambda}}{2}t} + Fe^{\frac{-\delta - \sqrt{\delta^2 - 4\lambda}}{2}t} - \frac{k}{\lambda} $$
$$ X = Ae^{Jx} + Be^{-Jx} + Ccos(Jx) + Dsin(Jx) $$
$$ u = XT $$

So, we now have the genreal solution. It is now time to apply boundary conditions to see if we can get rid of any of these terms. There are no special boundary conditions on T, so we can't assert anything. Since we suspect it will actually be oscillatory, we can't remove the "exploding" term either. 

There are conditions on x, however. The base function needs to be 0 at x=0 and x=$\pi$. *immediately* this removes the majority of our terms: A, B, and C's terms cannot meet this no matter what. Thus:

$$ X = Dsin(Jx) $$

The other boundary condition deals with the second derivative of this. 

$$ X'' = - DJ^2sin(Jx) $$

Which does in fact satisfy the second derivative being zero at 0 and $\pi$, but only for certain values of J. Actually, we know the values of this J already, this is the normal sine result, $\frac{n\pi}{L}$, which in this case reduces to simply n since L is $\pi$.

$$ X = Dsin(nx)$$

Note that this means $\lambda = n^4$, as J was the fourth root. Now, what values, exactly, does n take? Since we assumed positive and have already denied $\lambda=0$, we have 1 and higher. This also allows us to rewrite the equations for T.

$$ T = Ee^{\frac{-\delta + \sqrt{\delta^2 - 4n^4}}{2}t} + Fe^{\frac{-\delta - \sqrt{\delta^2 - 4n^4}}{2}t} - \frac{k}{n^4} $$

Now, if we go for small $\delta$, that is $\delta << 1$, then we rather clearly see that the terms in the exponents have to be complex, producing oscillatory results--however, it is oscillatory results that have an exponential effect *also* in place over them due to the real part.

$$ T = Ee^{-\delta t/2}cos(\frac{(-i)\sqrt{\delta^2-4n^2}}{2}t) + Fe^{-\delta t/2}sin((-i)\frac{\sqrt{\delta^2-4n^2}}{2}t) -\frac{k}{n^4} $$

If we commit to removing $\delta$ from the square roots, we end up with

$$ T = Ee^{-\delta t/2}cos(nt) + Fe^{-\delta t/2}sin(nt) -\frac{k}{n^4} $$

Note what would have hapened here if $\lambda$ were negative, the oscillatory portions would have gone away because the roots would not have been imaginary. This woudln't make physical sense, beams vibrate when they are disturbed, oscillating up and down until they reach their equilibrium point--this is reflected here, for as time goes on, the exponential portions dampen the oscillations within. 

So, now we have a single full solution:

$$ u_n(x,t) = \left[ Ee^{-\delta t/2}cos(nt) + Fe^{-\delta t/2}sin(nt) -\frac{Dk}{n^4} \right] sin(nx) $$

And the FULL solution would be:

$$ u(x,t) = \sum_{n=1}^\infty \left[ E_ne^{-\delta t/2}cos(nt) + F_ne^{-\delta t/2}sin(nt) -\frac{D_nk}{n^4} \right] sin(nx) $$

Note that we do not have expressions for $E_n, F_n$ or $D_n$. Also note that the constant D was able to be consolidated into other constants with E and F, but not on the constant term. 

A complete solution would find expressions for these constants. Unfortunately we don't have any boundary conditions that help with this, as the second derivative cares only about x, and the x portion of this result is certainly zero at x=0 and x=$\pi$ for both the normal function and the second derivative. 

Of course, the inability to fully define the constants should actually be expected, this is a *fourth order* differential equation. We have the prerequiside boundary conditions (four of them) to restrict everything, but we do not have *any* initial conditions provided to us. Furthermore, we suspect we would need multipel types of initial conditions, either at multiple locations or at multiple derivatives. All we can physically say is that the behavior should dampen over time which is built in to *all* of the time-dependent portions, so we can't get rid of any of it. So, alas, we must declare the question done at this point.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

This can be read right from the textbook on page 139, the natural frequenceis $\frac{n\pi c}{L}$ with the fundamental mode having a frequency of $\frac{\pi c}{L}$. Note that the constant c is $\sqrt{T_0/\rho_0}$, so the full answer would be

$$ \frac{n\pi}{L} \sqrt{\frac{T_0}{\rho_0}} $$

Now this requires solving the wave equation agian, though a good chunk of the steps are already done. We just replace the u(H,t)=0 condition with $u_x(H,t)=0$. The separation of variables remains the same, and since this is a condition that is a derivative with respect to x, the temporal portion (T) of the solution does not change. The X portion, however... the solution is one of sines and cosines. Since we need to be zero at x=0, we need a solution that takes the form of sines. 

$$ X = A sin(\sqrt{\lambda} x) $$

However, the other condition requries the *derivative* be zero. Meaning that our condition on $\lambda$ is not the same as usual; it means...

$$ X'(H) = A cos(\sqrt{\lambda} H) = 0 $$

The solution here is not $\left(\frac{n\pi}{H}\right)^2$ but rather $\left(\frac{n\pi - (\pi/2)}{H}\right)^2$. We have chosen $-\pi/2$ as the offset since that ensures n=1 is the first valid choice for the integer. 

Which would change T in a similar way. The end result would be:

$$ u(x,t) = \sum_{n=1}^\infty sin\frac{(n\pi - \pi/2) x}{H} \left( A_n cos\frac{(n\pi - \pi/2) ct}{H} + B_n sin\frac{(n\pi - \pi/2) ct}{H}\right) $$

We extract from this a frequency of

$$ \frac{n\pi - \pi/2}{H} \sqrt{\frac{T_0}{\rho_0}} $$

This is something that can be argued. If H=L/2, then we have to have "crests" of the sine function over the L/2 point for all harmonics part b). For the a) situation, we see that the crests *do* in fact exist over L/2 whenever the n number is odd. When n is even, the L/2 position has a *node* because the harmonic is able to split the length down the middle. It's basic symmetry: in theory, the "wave" of b) continues past L/2, and so it becomes the exact same wave as the full L. However, the number of these frequencies aren't likely to match up, since the a) n=2 case does not exist for the b) cases. But one case that *will* match is n=1.

$$ \frac{\pi - \pi/2}{L/2} \sqrt{\frac{T_0}{\rho_0}} = \frac{\pi}{L} \sqrt{\frac{T_0}{\rho_0}} $$

Which is in fact the first frequency for the a) situation. The second situaiton...

$$ \frac{2\pi - \pi/2}{L/2} \sqrt{\frac{T_0}{\rho_0}} = \frac{3\pi}{L} \sqrt{\frac{T_0}{\rho_0}} $$

Which would match up with the n=3 sitation in part a). Which it rather clearly does. In fact we could continue this forever, noting that the relationship is a rising correlation which can be written as a funciton: a = 2b-1, where a is the a) mode and b is the b) mode. And to prove the frequencies match in genreality...

$$ \frac{n\pi - \pi/2}{L/2} \sqrt{\frac{T_0}{\rho_0}} = \frac{(2n-1)\pi}{L} \sqrt{\frac{T_0}{\rho_0}} $$

Behold, the correlation is the same. With that, we are done.
