# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Part of us feels as though the fourier cosine series for x is a known result, but we shall treat it as though it is not, since we have recently been cautioned to show more steps. The cosine series comes from:

$$ f(x) \text{~} \sum_{n=0}^\infty A_n cos\frac{n \pi x}{L} $$
$$ A_0 = \frac1L \int_0^L f(x) dx $$
$$ A_n = \frac2L \int_0^L f(x) cos\frac{n\pi x}{L} $$

We shall treat this in genreality, so L can be any positive value. Our funciton is just x. 

$$ A_0 = \frac1L \int_0^L x dx = \frac1L \frac{L^2}{2} = \frac L2 $$

$$ A_n = \frac2L \int_0^L x cos\frac{n\pi x}{L} $$

Explicitly integrate by parts for thoroughness. 

$$ = \frac2L \left[ \frac{L}{n\pi} xsin\frac{n\pi x}{L}|_0^L- \int_0^L \frac{L}{n\pi}sin\frac{n\pi x}{L} \right] $$
$$ = \frac2L \left[ \frac{L^2}{n^2\pi^2} cos\frac{n\pi x}{L} \right]_0^L $$
$$ = \frac2L \left[ \frac{L^2}{n^2\pi^2} cos(n\pi) - \frac{L^2}{n^2\pi^2} \right] $$
$$ = \frac{2L}{n^2\pi^2} \left[ (-1)^n - 1 \right] $$

And so...

$$ x \text{~} \frac{L}{2} + \sum_{n=1}^\infty \frac{2L}{n^2\pi^2} \left[ (-1)^n - 1 \right]  cos\frac{n\pi x}{L}$$

Which does produce the cosine series for x. Here it is for L=1.



From this alone, we can learn a few things. First of all, if we set x=0, we get the following series:

$$ 0 = \frac{L}{2} + \sum_{n=1,3,5...}^\infty \frac{-4L}{n^2\pi^2} $$

From which we can extract

$$ \Rightarrow \frac L2 = \frac{4L}{\pi^2}\sum_{n=1,3,5...}^\infty \frac{1}{n^2} $$
$$ \Rightarrow \frac{\pi^2}{8} = \sum_{n=1,3,5...}^\infty \frac{1}{n^2} $$

Which is precisely half of what we need, as it only adds up the odd terms. We need some way to add up the even terms, and then we can combine both into the total series. 

Now, something about the even series and the total series. The even series could be written as $\frac{1}{4n^2}$, or exactly 1/4th the total series. We can treat this as a variable, meaning Total Series = Odd + Even series becomes Total = Odd + Total/4, or Odd = 3Total/4, or... Total = 4Odd/3. 

Which means...

$$ \frac43 \frac{\pi^2}{8} = \frac{\pi^2}{6} = \sum_{n=1}^\infty \frac{1}{n^2} $$

Which is the series we sought to solve.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

Let's review the complex Fourier series. 

$$ f(x) \text{~} \sum_{n=-\infty}^\infty C_n e^{-in\pi x / L} $$
$$ C_n = \frac{1}{2L} \int_{-L}^{L} f(x) e^{in\pi x / L} dx $$

This proof can be accomplished by evaluating the integral in generality for both +n and -n. We need to show that one is the complex conjugate of the other. Fortunately, there is only one change in the two integrals, a single sign change on n which is identical to changing the sign of i. 

$$ \frac{1}{2L} \int_{-L}^{L} f(x) e^{in\pi x / L} dx ; \frac{1}{2L} \int_{-L}^{L} f(x) e^{-in\pi x / L} dx $$

Of cousre, it is conceivable that the interior of an integral doesn't simply complex conjugate like that, so we have to show the relation in a more explicit manner. 

Integration by parts revelead nothing, but there is something we can do: split the exponential up into trig components. 

$$ \frac{1}{2L} \int_{-L}^{L} f(x) cos\frac{n \pi x}{L} dx + \frac{1}{2L} \int_{-L}^{L} f(x) i sin\frac{n \pi x}{L} dx   ; \frac{1}{2L} \int_{-L}^{L} f(x) cos\frac{n \pi x}{L} dx - \frac{1}{2L} \int_{-L}^{L} f(x) i sin\frac{n \pi x}{L} dx $$

Now we can just, rather trivially, pull the "i" out for it is a constant.

$$ \frac{1}{2L} \int_{-L}^{L} f(x) cos\frac{n \pi x}{L} dx + i \frac{1}{2L} \int_{-L}^{L} f(x) sin\frac{n \pi x}{L} dx   ; \frac{1}{2L} \int_{-L}^{L} f(x) cos\frac{n \pi x}{L} dx - i \frac{1}{2L} \int_{-L}^{L} f(x) sin\frac{n \pi x}{L} dx $$

Now, every integral in these two formulations are definite integrals, which means they go to known constants. The integrals are identical between both terms. In fact, the only difference is the sign of that i, but now that the i is outside the integral, we know we can take the complex conjugate directly just by flipping its sign. Thus, $C_{-n} = \bar{C_n}$

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

The separation of variables solution, as outlined in the book, takes the form 4.4.11:

$$ u(x,t) = \sum_{n=1}^\infty \left( A_n sin\frac{n\pi x}{L}cos\frac{n\pi ct}{L} + B_n sin\frac{n\pi x}{L} sin\frac{n \pi ct}{L}\right) $$

This, notably, is for when u(x,0) = f(x). The actual coefficients are given by 4.4.13:

$$ A_n = \frac 2L \int_0^L f(x) sin\frac{n\pi x}{L} dx $$
$$ B_n \frac{n \pi c}{L} = \frac 2L \int_0^L g(x) sin\frac{n\pi x}{L} dx $$

Anyway, with f(x) = 0, the $A_n$ coefficients also go to zero, changing the result to:

$$ u(x,t) = \sum_{n=1}^\infty \left( B_n sin\frac{n\pi x}{L} sin\frac{n \pi ct}{L}\right) $$

Now here's where the actual problem arises: how do we turn an infinite sum into an integral? A curious puzzle, especially because we don't actually know what g(x) is. But let's perform the full substitution so we have everything to work with that we can.

$$ u(x,t) = \sum_{n=1}^\infty \left( \frac{L}{n \pi c}\frac 2L \int_0^L g(x) sin\frac{n\pi x}{L} dx sin\frac{n\pi x}{L} sin\frac{n \pi ct}{L}\right) $$
$$ = \frac{2}{n \pi c} \sum_{n=1}^\infty \left( \int_0^L g(x) sin\frac{n\pi x}{L} dx sin\frac{n\pi x}{L} sin\frac{n \pi ct}{L}\right) $$

Now before we start jumping around pulling sums in and out of integrals and vice-versa, let's try to look at what our goal is. How would we even get G(x) out of this? Well, G(x), being the odd extension of g(x), can be obtianed through the fourier sine series. Which is to say:

$$ G(x) = \sum_{n=1}^\infty \frac2L \int_0^L g(x) sin\frac{n \pi x}{L} dx sin\frac{n \pi x}{L} $$

Which is almost what we have in our relation... EXCEPT for the ct term right there at the end. So, clearly, we're going to have to find a way to work around that.  

One thing we can do, though, is use our knoweldge of trig functions (and 4.4.14) to make the variable not x and ct, but x+ct and x-ct. 

$$ u(x,t) = \frac{2}{n \pi c} \sum_{n=1}^\infty \left( \int_0^L g(x) sin\frac{n\pi x}{L} dx \left[ \frac12 cos\frac{n\pi(x-ct)}{L} - \frac12 cos\frac{n\pi(x+ct)}{L}\right]\right) $$ 

Now, doesn't the term in brackets... look kind of like the result of a definite integral? From, say, x-ct to x+ct? Let this mysterious "original" variable be $\xi$. 

$$ u(x,t) = \frac{2}{n \pi c} \sum_{n=1}^\infty \left( \int_0^L g(x) sin\frac{n\pi x}{L} dx \int_{x+ct}^{x-ct} -\frac{n\pi}{L 2}sin\frac{n\pi \xi}{L} d\xi\right) $$ 

Flip the sign by flipping the bounds.

$$ u(x,t) = \frac{1}{2 c} \sum_{n=1}^\infty \left( \frac 2L \int_0^L g(x) sin\frac{n\pi x}{L} dx \int_{x-ct}^{x+ct} sin\frac{n\pi \xi}{L} d\xi\right) $$ 


Now, the thing we need to realize is that our "integral over dx" is actually arbitrary, we can transform it into an "integral over d$\xi$" and nothing changes since the result is a number. As it is a number, it can be pulled inside the other integral. The result?

$$ u(x,t) = \frac{1}{2 c} \sum_{n=1}^\infty \left( \int_{x-ct}^{x+ct} \frac 2L \int_0^L g(\xi) sin\frac{n\pi x}{L} d\xi sin\frac{n\pi \xi}{L} d\xi\right) $$ 

And, as we know, sums and integrals are interchangeable.

$$ u(x,t) = \frac{1}{2 c} \left( \int_{x-ct}^{x+ct} \sum_{n=1}^\infty  \frac 2L \int_0^L g(\xi) sin\frac{n\pi x}{L} d\xi sin\frac{n\pi \xi}{L} d\xi\right) $$ 

And we now have the definition we outlined for G(x). Technically only valid on the bounds -L to L, but that's all we need for this problem.

$$ u(x,t) = \frac{1}{2 c}  \int_{x-ct}^{x+ct} \sum_{n=1}^\infty  G(\xi)  d\xi$$ 

Which is what we were going for. The leading constant is right, the sign is right, and the bounds are even right. Problem complete!

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

The separation of variables solution, as outlined in the book, takes the form 4.4.11:

$$ u(x,t) = \sum_{n=1}^\infty \left( A_n sin\frac{n\pi x}{L}cos\frac{n\pi ct}{L} + B_n sin\frac{n\pi x}{L} sin\frac{n \pi ct}{L}\right) $$

This, notably, is for when u(x,0) = f(x). The actual coefficients are given by 4.4.13:

$$ A_n = \frac 2L \int_0^L f(x) sin\frac{n\pi x}{L} dx $$
$$ B_n \frac{n \pi c}{L} = \frac 2L \int_0^L g(x) sin\frac{n\pi x}{L} dx $$

We will assume that we are provided an f(x) and g(x), but we will solve in generality for any possible solution with f(x) and g(x). It should hold for all of them. 

So, fortunately, $A_n$ and $B_n$ are constants and thus can be treated as such when we do things such as integrating. Anyway, to show that E is constant in time, we need to find E in general and show that adjusting t does not change the result. To do that, we need to take some basic derivatives. One with respect to t, and one with respect to x.

$$ u_t(x,t) = \sum_{n=1}^\infty \left( -A_n\frac{n\pi c}{L} sin\frac{n\pi x}{L}sin\frac{n\pi ct}{L} + B_n \frac{n\pi c}{L} sin\frac{n\pi x}{L} cos\frac{n \pi ct}{L}\right) $$
$$ u_x(x,t) = \sum_{n=1}^\infty \left( A_n \frac{n\pi}{L}cos\frac{n\pi x}{L}cos\frac{n\pi ct}{L} + B_n \frac{n\pi}{L} cos\frac{n\pi x}{L} sin\frac{n \pi ct}{L}\right) $$

Note that we are assuming term-by-term differentiation works here. Which is reasonable, since any physically real solution to the wave equation has to be a continuous wave. 


The next thing we want to do is square these series, but that becomes a huge can of worms very quickly. However, we can make use of orthogonality to get a result out of this. The integrals we are taking in the energy equation are over x, so all the "t" terms might as well be constants now that we've already taken the derivatives. To make our orthogonality relations easier to see, we will temporarily consolidate these into the coefficients.

$$ u_t(x,t) = \sum_{n=1}^\infty \left( -A_n' sin\frac{n\pi x}{L}+ B_n' sin\frac{n\pi x}{L} \right) $$
$$ u_x(x,t) = \sum_{n=1}^\infty \left( A_n'' cos\frac{n\pi x}{L} + B_n''  cos\frac{n\pi x}{L} \right) $$

Now, when squaring, every possible value of n will be paired with every other possible value of n. However, we know from the ortogonality relations that only when the n values are equal will anything come out of it. (And since n cannot be zero, we don't have to worry about anything unusual, either). Thus we can actually write down the expression for E.

$$ \int_0^L \frac12 \sum_{n=1}^\infty \left[ \frac{n^2\pi^2 c^2}{L^2}sin^2\frac{n \pi x}{L}\left( B_ncos\frac{n\pi ct}{L} - A_nsin\frac{n\pi ct}{L}  \right)^2 \right] dx + \int_0^L \frac{c^2}{2} \sum_{n=1}^\infty \left[ \frac{n^2\pi^2}{L^2}cos^2\frac{n \pi x}{L}\left( A_ncos\frac{n\pi ct}{L} + B_nsin\frac{n\pi ct}{L}  \right)^2 \right] dx$$

Note that the sum still exists, it's just that only the same n (or "diagonal") terms that still exist. Another property we make use of ist he ability to shuffle sums and integrals. Evaluating nothing, just using shuffling, we arrive at a single extended sum:

$$  \sum_{n=1}^\infty \left[  \frac{n^2\pi^2 c^2}{2L^2} \left( B_ncos\frac{n\pi ct}{L} - A_nsin\frac{n\pi ct}{L}  \right)^2 \int_0^Lsin^2\frac{n \pi x}{L} dx+  \frac{n^2\pi^2 c^2}{2L^2} \left( A_ncos\frac{n\pi ct}{L} + B_nsin\frac{n\pi ct}{L}  \right)^2 \int_0^L cos^2\frac{n \pi x}{L} dx\right]$$

The integral of that square cosine is just L/2, as seen in 2.4.22. In previous homeworks we've also done the square sine, which is also L/2. So we now have...

$$  \sum_{n=1}^\infty \left[  \frac{n^2\pi^2 c^2}{4L} \left( B_ncos\frac{n\pi ct}{L} - A_nsin\frac{n\pi ct}{L}  \right)^2 +  \frac{n^2\pi^2 c^2}{4L} \left( A_ncos\frac{n\pi ct}{L} + B_nsin\frac{n\pi ct}{L}  \right)^2 \right]$$

Now, this looks like it may be t dependent, but it really isn't since the term on the left is precisely 180 degrees out of phase with the term on the right, but the constants make this so it isn't self-evident. And we should work it out anyway. REgardless, since the leading constants are the same, we can remove a lot of superfluous information, leading us to only evaluate

$$ \left( B_ncos\frac{n\pi ct}{L} - A_nsin\frac{n\pi ct}{L}  \right)^2 +  \left( A_ncos\frac{n\pi ct}{L} + B_nsin\frac{n\pi ct}{L}  \right)^2 $$

for arbitrary n. 

$$ = B_n^2cos^2\frac{n\pi ct}{L} + A_n^2sin^2\frac{n\pi ct}{L} - 2A_nB_nsin\frac{n\pi ct}{L}cos\frac{n\pi ct}{L} + A_n^2cos^2\frac{n\pi ct}{L} + B_n^2sin^2\frac{n\pi ct}{L} + 2A_nB_nsin\frac{n\pi ct}{L}cos\frac{n\pi ct}{L} $$
$$ = B_n^2 + A_n^2 $$

Which is a constant, that is, independent of time. So we have shown what we sought. Technically we have only shown that each individual term is independent of t in the sum, but if all terms are independent of t, then the sum is as well. 

Curiously, we have essentially already done this! An individual mode is a solution for one value of n, say, n=1 or n=2. n=1 is a special case called the normal mode. When we wrote the energy out in full in part a), it was in the form of the sum. But since we only cared about proving t-dependence, we removed the sum and only considered one term. That term is, in fact

$$ E_n \propto B_n^2 + A_n^2 $$

That is, the energy of a single mode is proportional to the sum of the square terms of the Fourier Series. The textbook also notes that this is the square amplitude, which makes sense. If we want the actual answer, however, we need to go back a step in part a) and add back the leading constant.

$$ E_n = \frac{n^2\pi^2c^2}{4L}(B_n^2 + A_n^2) $$

WHich is our answer. 


This, too, we essentially already did in part a) (which makes us think that there was a far easier way of showing t-independence than the fully explicit derivation we did.) If we go back a few steps and *don't* remove the sum:

$$ E = \sum_{n=1}^\infty \left[  \frac{n^2\pi^2 c^2}{4L} \left( B_ncos\frac{n\pi ct}{L} - A_nsin\frac{n\pi ct}{L}  \right)^2 +  \frac{n^2\pi^2 c^2}{4L} \left( A_ncos\frac{n\pi ct}{L} + B_nsin\frac{n\pi ct}{L}  \right)^2 \right]$$

We can still consolidate all the t terms.

$$ E = \sum_{n=1}^\infty \left[  \frac{n^2\pi^2 c^2}{4L} (A_n^2 + B_n^2)\right]$$

Which has the energy be the sum of the energy of each individual mode, exactly what we sought. 

A note: mode does not mean node, and it is easay to get this confused. However, the energy between two nodes also adds to the total energy, so b) and c) still hold true for nodes. Granted, nodes exist as subdivisions of a mode, so it's just chopping it up into smaller pieces.
