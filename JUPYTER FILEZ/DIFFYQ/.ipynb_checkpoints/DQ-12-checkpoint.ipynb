# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

This is a little awkward, as the initial condition doesn't seem physically real: the boundary has to be 0 everywere on the boundary, but the temperature itself needs to start at 100 everywhere, which indicates a sharp discontinuity, which is kind of against the whole point. Since we don't specify that heat cannot flow out, eventually as time goes to infinity naturally all the heat must flow out into the resevoir of zero and approach a total temperature of zero. This is assuming the resevoir is large, otherwise the small "boundary" of 0 temperature does nothing to the temperature inside. 

We can split up the function u into two parts: A and T, for area and time. A is a function of r and $\theta$, while T is a function of T. We assume $\kappa$ is a constant. This makes the separation of variables:

$$ AT_t = \kappa T\nabla^2 A $$
$$ \frac{1}{T\kappa} T_t =  \frac{1}{A} \nabla^2 A $$

Which gives us the separated equations:

$$ T_t + \lambda \kappa T = 0 $$
$$ \nabla^2 A + \lambda A = 0 $$

The time solution is simple: $ T = e^{-\kappa \lambda t} $ with some leading constant. 

Now the physical solution is a bit differnet but it IS in a known form. 



Specifically, we can consider it a S-L form with p=$\sigma$=1 and q=0. Which we can then separate into radial and polar parts, following the forms on 7.7.12 and 7.7.13

$$ \Theta_{\theta\theta} + \mu \Theta = 0 $$
$$ r (r R_r)_r + (\lambda r^2 - \mu)R = 0 $$

Due to the nature of circles, $\Theta$ must be periodic. This is a problem that's been done before, we know the eigenvalue $\mu=m^2$ and that our options are $sinm\theta$ and $cosm\theta$ for the eigenfunction. This adjusts our radikal function slightly.

$$ r (r R_r)_r + (\lambda r^2 - m^2)R = 0 $$

And we then expand the derivative...

$$ r^2 R_{rr} + rR_r + (\lambda r^2 - m^2)R = 0 $$

We can remove the eigenvalue dependence by performing a substitution.

$$ z^2 R_{zz} + zR_z + (z^2 - m^2)R = 0 $$

This is Bessel's Differential Equation of order m. The general solution is then:

$$ R = C_1 J_m(z) + C_2 Y_m(z) $$

Now, our situation involves the center, so we need it to be finite at r=0, which Y does not satisfy.

And the condition $J_m(\sqrt\lambda a) = 0$ will give us the specific eigenvalues, but those will be tabulated and not our concern at the moment. 

And so we can actually write out the full solution.

$$ u(r,\theta,t) = \sum_{m=0}^\infty \sum_{n=1}^\infty A_{mn} J_m(\sqrt{\lambda_{mn}}r) cos(m\theta) e^{-\kappa \lambda_{mn} t} + \sum_{m=1}^\infty \sum_{n=1}^\infty B_{mn} J_m(\sqrt{\lambda_{mn}}r) sin(m\theta) e^{-\kappa \lambda_{mn} t} $$

Which is the solution without any initial conditions to specify A and B. At this point we can actually answer part of the problem: comment on the behavior at large t. This makes the exponential time-dependend terms go to zero, making the entire temperature go to zero. This makes sense, we are not dictating that heat flow in must equal heat flow out--naturally a hot object in a sea of 0-degree something will give up all its heat to that 0-degree something. It will eventually go to nothing. 

However, we were asked for the full solution. So we have to take the initial condition and find A and B. (Though since we're using a Bessel Function we have doubts that there will be a simple closed-form answer.)



Once we consider initial conditions, we actually have a very good simplificaiton. Our initial condition is 100 everywhere, which is independent of $\theta$ so we have circular symmetry, and naturally with no outside forces the solution must remain angularly symmetric as well. The only thing we can set to make the angular appearance always vanish is m=0, which will set our sine term to zero. This means we can simplify...

$$ u(r,\theta,t) = \sum_{n=1}^\infty A_{n} J_0(\sqrt{\lambda_{n}}r) e^{-\kappa \lambda_{n} t} $$

Still very clearly decays at large t. 

Let's set t=0 and use the initial conditions. 

$$ u(r,\theta,0) = 100 = \sum_{n=1}^\infty A_{n} J_0(\sqrt{\lambda_{n}}r) $$

Following 7.7.66, we twist this into a form to find $A_n$

$$ A_n = \frac{\int_0^a 100 J_0(\sqrt{\lambda_{n}}r) r dr}{\int_0^a J_0^2(\sqrt{\lambda_{n}}r) r dr} $$

And those are integrals we cannot evaluate in generality. So we consider ourselves done, with the solution:

$$ u(r,\theta,t) = \sum_{n=1}^\infty A_{n} J_0(\sqrt{\lambda_{n}}r) e^{-\kappa \lambda_{n} t} $$
$$ A_n = \frac{\int_0^a 100 J_0(\sqrt{\lambda_{n}}r) r dr}{\int_0^a J_0^2(\sqrt{\lambda_{n}}r) r dr} $$
$$ J_0(\sqrt\lambda_n a) = 0 $$

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

Let $\phi$ = XY and separate the variables.

$$ YX_{xx} + XY_{yy} + XY \lambda = 0 $$
$$ \Rightarrow \frac1X X_{xx} + \frac1Y Y_{yy} + \lambda = 0 $$

Looks like we'll have to introduce a new constant, let it be $\mu$. Now we have:

$$ \Rightarrow X_{xx} + X \mu = 0 $$
$$ \Rightarrow -Y_{yy} + Y (-\lambda + \mu) = 0 $$

Which are equations that are similar in form, but not identical. Also it would be trivial to replace $\lambda+\mu$ with just $\mu$ and vice versa if it would be necessary. 

The X solution is sines and cosines. 

$$ X = Asin(\sqrt{\mu}x) + Bcos(\sqrt{\mu}x) $$

The Y solution can be represented similarly

$$ Y = Csin(\sqrt{-\mu+\lambda}y) + Dcos(\sqrt{-\mu+\lambda}y) $$


We need zero at the edges, so everything must be a sine function.

$$ X = Asin(\sqrt{\mu}x) $$

$$ Y = Csin(\sqrt{-\mu+\lambda}y) $$

Now, restricting $\mu$ is easy, we've done it a million times, $\mu = (\frac{n\pi}{L})^2 = n^2$ where n is some integer starting at 1. 

$$ X = Asin(n x) $$

$$ Y = Csin(\sqrt{-n^2+\lambda}y) $$

Now this *also* restricts $\lambda$ to taking a set number of values. It needs to very specifically make the result of the square root be a whole number, and a whole positive number at that due to the square root. This means $\lambda = n^2 + m^2$, where m is some other constant. There's no restriction that requries x and y to be oscillating at the same speed, after all, we could combine things in multiple ways. 

$$ X = Asin(n x) $$

$$ Y = Csin(m y) $$

And so in total we have:

$$ \phi(x,y) = \sum_{nm} A_{nm} sin(nx)sin(my) $$

The eigenvalues themselves were $\lambda = n^2 + m^2$ which came from $\mu = (\frac{n\pi}{L})^2 = n^2$

As we have no initial conditions, this is as far as we can go. 

The assumption we are making is essentially that the answer to the Helmhloltz and Poisson equation have the same form. 

$$ u(x,y) = \sum_{nm} A_{nm} sin(nx)sin(my) $$

Now we get to shunt this INTO the original PDE to see what happens. 

$$ (\sum_{nm} A_{nm} sin(nx)sin(my))_{xx} + (\sum_{nm} A_{nm} sin(nx)sin(my))_{yy} = Q(x,y) $$

Term by term differentiation is just fine. 

$$ \Rightarrow - n^2 (\sum_{nm} A_{nm} sin(nx)sin(my)) - m^2 (\sum_{nm} A_{nm} sin(nx)sin(my)) = Q(x,y) $$

We can multiply through by sin(n'x)sin(m'y) and then integrate, abusing orthogogonality over both x and y. 

$$ \Rightarrow - n^2 (\sum_{nm} A_{nm} sin(nx)sin(my)sin(n'x)sin(m'y)) - m^2 (\sum_{nm} A_{nm} sin(nx)sin(my)sin(n'x)sin(m'y)) = Q(x,y)sin(n'x)sin(m'y) $$

The results of the double integral are given to us, and the orthogonality makes everything else go to zero. (And we did check, the integral of $sin^2(nx)$ is $\pi/2$ and so naturally the total over x and y is $\pi^2/4$. 

$$ \Rightarrow - n^2 A_{nm} \frac{\pi^2}{4} - m^2 A_{nm} \frac{\pi^2}{4} = \int_0^\pi \int_0^\pi Q(x,y)sin(nx)sin(my) dxdy $$
$$ \Rightarrow A_{nm} = \frac{4 \int_0^\pi \int_0^\pi Q(x,y)sin(nx)sin(my) dxdy}{\pi^2 (-n^2-m^2)} $$

And since Q is of an unknown form we cannot evaluate that integral, but if we did know Q we could get an explicit result.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

consider u to be composed of XYT. From that we can quickly extract:

$$\frac{1}{T} T_t = \frac{1}{XY} (XY)_{xx} + \frac{1}{XY} (XY)_{yy} + 4 $$

Which becomes

$$ T_t + \lambda T = 0 $$
$$ YX_{xx} + XY_{yy} + 4XY + \lambda XY $$

And we can split the second one up further. 

$$ X_{xx} +\mu X = 0 $$
$$ -Y_{xx} - 4Y - \lambda Y + \mu Y = -Y_{yy}-(4+\lambda-\mu)Y = 0  $$

We have now separated the variables. 

The time component is pretty clearly $e^{-\lambda t}$, exponential decay. 

The X one would then be $cos(\sqrt\mu x)$. This also gives us strict limitaitons on $\mu: (\frac{n \pi}{L})^2 = n^2\pi^2$ (Cosine required due to boundary conditions being on the derivative of x, not x itself.)

Now we have to be a bit more careful with what form Y takes. The solution is trigonometric.

$$ Y = Acos\sqrt{4+\lambda-\mu}y + Bsin(\sqrt{4+\lambda-\mu}y) $$

Only sine functions survive, as we need it to be zero at y=0. 

$$ \Rightarrow Y = sin(\sqrt{4+\lambda-\mu}y) $$

Now we have $\mu$ already.

$$ \Rightarrow Y = sin(\sqrt{4+\lambda-n^2\pi^2}y) $$

Now, the *second* boundary condition usually has us set this to zero. However, we can't do that in this case, as the boundary condition is $Y'(1) + Y(1) = 0$, rather than just a quick boundary, so we have to actually take the derivative to see. 

$$ \sqrt{4+\lambda-n^2\pi^2}cos(\sqrt{4+\lambda-n^2\pi^2}) + sin(\sqrt{4+\lambda-n^2\pi^2}) = 0 $$ 

Which we can rearrange to

$$ \sqrt{4+\lambda-n^2\pi^2} = - tan(\sqrt{4+\lambda-n^2\pi^2}) $$ 

And to avoid writing things out in annoying long form, we can set the interior to a different variable, one dependent on *m* rather than n since, when we solve for this, the n dependence in $\lambda$ is there just to get rid of the n term on the other side. (Or it can at least be arranged in such a way).

$$ \sqrt{\mu_m} = - tan(\sqrt{\mu_m}) $$

As this is a transcendedntal equation we shall go no further. 

This would make our full set of equations be:

$$ T = e^{-\lambda_{mn} t}$$
$$ X = cos(n\pi x) $$
$$ Y = sin(\sqrt{\mu_m}y) $$

$$ u(x,y,t) = \sum_{nm} A_{nm} cos(n\pi x) sin(\sqrt{\mu_m}y) e^{-\lambda_{mn} t} $$

Since we have an initial condition we can actually define the coefficients A. Set t=0 and use the initial conditions. 

$$ \alpha(x,y) = \sum_{nm} A_{nm} cos(n\pi x) sin(\sqrt{\mu_m}y)$$

Now, that *last* y-related part is proving problematic, as it's not obvious exactly how it is orthogonal and in what way. However, as it is a sine function of the form sin(ax) we know it has an orthogonality set based on the values of $\lambda$. The issue is determining the precise value of said orthogonality integral. Let's just see where we end up...

$$ \Rightarrow \alpha(x,y) = \sum_{nm} A_{nm} cos(n\pi x) sin(\sqrt{\mu_m}y)$$

Now we can do the orthogonality integral that we're familiar with at this point in time. 

$$ \Rightarrow \int\int\alpha(x,y) cos(n\pi x) sin(\sqrt{\mu_m}y)dxdy =  \sum_m A_{nm} \pi/2 \int sin(\sqrt{\mu_m}y) sin(\sqrt{\mu_{m'}}y) dy $$
$$ \Rightarrow A_{nm} = \frac{2 \int\int\alpha(x,y) cos(n\pi x) sin(\sqrt{\mu_m}y) dxdy }{\pi \int sin^2(\sqrt{\mu_m}y) dy} $$

The last step can only be done since we know the functions have to be orthogonal with themselves, all eigenfunctions and eigenvalues must obey that. With that, however, we CAN evaluate the lower integral since all the internal stuff is just a constant. 

$$ \Rightarrow A_{nm} = \frac{2 \int\int\alpha(x,y) cos(n\pi x) sin(\sqrt{\mu_m}y) dxdy }{\pi \left( -\frac{sin(2\sqrt\mu_my)}{4\sqrt{\mu_m}} + \frac{y}{2} \right)|_0^1} $$

$$ \Rightarrow A_{nm} = \frac{2 \int\int\alpha(x,y) cos(n\pi x) sin(\sqrt{\mu_m}y) dxdy }{\pi \left( -\frac{sin(2\sqrt\mu_m)}{4\sqrt{\mu_m}} + \frac{1}{2} \right)} $$

$$ \Rightarrow A_{nm} = \frac{2 \int\int\alpha(x,y) cos(n\pi x) sin(\sqrt{\mu_m}y) dxdy }{\pi \left( \frac{2\sqrt{\mu_m} - sin(2\sqrt{\mu_m})}{4\sqrt{\mu_m}} \right)} $$

$$ \Rightarrow A_{nm} = \frac{4\sqrt{\mu_m} \int\int\alpha(x,y) cos(n\pi x) sin(\sqrt{\mu_m}y) dxdy }{\pi (2\sqrt{\mu_m} - sin(2\sqrt{\mu_m}))} $$

Which doesn't look to simplify much further. While there might be something to do with the double angle formula, it is likely to be of minimal simplificaiton. So we call this complete. With

$$ u(x,y,t) = \sum_{nm} A_{nm} cos(n\pi x) sin(\sqrt{\mu_m}y) e^{-\lambda_{mn} t} $$

And 

$$ \sqrt{\mu_m} = - tan(\sqrt{\mu_m}) $$
$$ \mu_m = 4+\lambda_{mn}-n^2\pi^2 $$

Completes our answer. 

$$ T = e^{-\lambda_{mn} t}$$

Now the obvious answer is "exponential decay" but... is it really? See, the solution we came up for $\lambda$ could be positive, negative, or even zero. Let's get a quick range on possible values:

$$\sqrt{4+\lambda-n^2\pi^2}$$ can never be negative so the minimum value of $\lambda$ is $n^2\pi^2 - 4$. When n=1 this is still positive, though, so this means that $\lambda$ is always positive! (and we know $\mu$ is positive due to the restrictions of X being a sine function). 

Since $\lambda$ has to be positive, everything exhibits exponential decay to nothing. 

This is much better in terms of the restriction of the Rayleigh Quotient, which can only show that $4+\lambda$ is positive. 

$$ \nabla^2 \phi + (4+\lambda) \phi = 0 $$

$$ 4+\lambda = \frac{-\oint \phi \nabla \phi \cdot \hat n dx + \int\int_R |\nabla\phi|^2 dxdy}{\int\int_R \phi^2 dx dy} $$

Since we have zero at the boundary, the boundary term goes away. 

$$ 4+\lambda = \frac{ \int\int_R |\nabla\phi|^2 dxdy}{\int\int_R \phi^2 dx dy} $$

But the two conclusions agree, so that's good. In conclusion, exponential decay at large times. This makes sense, we have a heat equation problem, it normally decays. The presence of a source Q would make it level off, if that Q did not depend on u itself. As u distributes itself, the strength of Q goes down. The rate of decay will be slowed by the source, naturally, but not removed entirely. It still seeks 0. 

$$ \Rightarrow A_{nm} = \frac{4\sqrt{\mu_m} \int\int\alpha(x,y) cos(n\pi x) sin(\sqrt{\mu_m}y) dxdy }{\pi (2\sqrt{\mu_m} - sin(2\sqrt{\mu_m}))} $$

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

Here we go again: let u = RT, for both radial and temporal situations, but we are physically in a system with theta coordinates existing, so we must be careful. Anyway, separation of variables:

$$RT_{tt} + 2bRT_t = \frac{1}{r} T(rR_r)_r $$
$$\frac1TT_{tt} + \frac2TbT_t = \frac{1}{rR}(rR_r)_r $$

And so

$$ T_{tt} + 2bT_t + \lambda T = 0 $$
$$ \frac{1}{r} (rR_r)_r +\lambda R = 0 $$

These is our separeted equations. We can expand the radial one a bit.

$$ R_{rr} + \frac{1}{r}R_r + \lambda R = 0 $$

Both equations are of similar form. Note that the radial one is just $\nabla^2 R + \lambda R = 0$. With the conditions of zero at the edge and finite interior, we have the exact situation as we do in 7.7.55. The full solutions are the bessel functions:

$$ R = AJ_0(\sqrt\lambda r) + BY_0(\sqrt\lambda r)$$

The zero is chosen because the solution differential equation is considered zeroth order among bessel funcitons. Y does not function as it has a sigularity at the origin. Thus:

$$ R = J_0(\sqrt\lambda r) $$

Now, the T situation is inherently different than the R situation, it's coefficients have no variables in them. However, this means we can use more general solving techniques. We can still construct the auxillary polynomial. 

$$ r^2 + 2br + \lambda $$

This is a situation of the quadratic formula, we can just evaluate the zeroes.

$$ = (r-b+\sqrt{b^2-\lambda})(r-b-\sqrt{b^2-\lambda}) $$



Which means that the temporal solution is:

$$ T = Ce^{(-b+\sqrt{b^2-\lambda})t} + De^{(-b-\sqrt{b^2-\lambda})t} $$

Now, we do have a limitation here: since we can have no negative square roots, $\lambda < b^2$. This means that the square root term is always of magnitude b or less, meaning the exponent is always negative, no matter what. Thus, both exponentials are perfectly valid and we can't just up and remove one. 

We can play with the initial conditions to narrow this down further. First of all, at t=0, we end up with:

$$T(0) = C+D$$

However, this needs to equal zero, so C+D=0 meaning they must be equal and opposite in terms of the actual constant. Meaning we can absorb it into one constant and leave the function with just its base form.

$$ T = e^{(-b+\sqrt{b^2-\lambda})t} - e^{(-b-\sqrt{b^2-\lambda})t} $$

We can actually simplify this a bit. 

$$ \Rightarrow T = e^{-bt}e^{(\sqrt{b^2-\lambda})t} - e^{-bt}e^{(-\sqrt{b^2-\lambda})t} $$

Which we can convert to sinh form. 

$$ \Rightarrow T = e^{-bt}sinh((\sqrt{b^2-\lambda})t) $$

And so the total combination is...

$$ u(r,t) = \sum_n A_n J_0(\sqrt\lambda_n r) e^{-bt}sinh((\sqrt{b^2-\lambda_n})t) $$

We actually have a limited number of eigenvalues! We know we must have positive values, such is the way things are with the bessel functions. We also know $\lambda < b^2$, so there would be a finite number of values n could take! However, there's not really a way to write them down, since the specific restriction is:

$$ J_0 (\sqrt\lambda_n a) = 0 $$

So we *can* find the eigenvalues from here, so we now claim we know them. However, we're looking for full solution, so let's find the constant. Which we are only going to be able to do with the t-derivative, since at t=0 everything is zero and we learn nothing. 

$$ u(r,t)_t = \sum_n A_n J_0(\sqrt\lambda_n r) \left[ (-b) e^{-bt}sinh((\sqrt{b^2-\lambda_n})t) + \sqrt{b^2-\lambda_n}) e^{-bt}cosh((\sqrt{b^2-\lambda_n})t) \right]$$

Now, at t=0...

$$ v_0 = \sum_n A_n J_0(\sqrt\lambda_n r) \left[ \sqrt{b^2-\lambda_n} \right]$$


Multiply through by $J_m$ and integrate, abusing orthogonality to remove the sum.

$$ \int_0^a v_0 J_0(\sqrt\lambda_n r) r dr = A_n \left[ \sqrt{b^2-\lambda_n} \right] \int_0^a J^2_0(\sqrt\lambda_n r) rdr$$

$$ A_n = \frac{v_0 \int_0^a J_0(\sqrt\lambda_n r) r dr}{\left[ \sqrt{b^2-\lambda_n} \right] \int_0^a J^2_0(\sqrt\lambda_n r) rdr} $$

Which we consider our answer. 

Now this we can examine looking just at T. 

$$ T = e^{(-b+\sqrt{b^2-\lambda})t} - e^{(-b-\sqrt{b^2-\lambda})t} $$
$$ T = e^{-bt}sinh((\sqrt{b^2-\lambda})t) $$

What we have here is two forms of the same function. It is not easy to see on the second (condensed) version, but on the exponential version it is clear to see: the exponent is always negative with the restrictions placed on $\lambda$ of being positive itself and being less than $b^2$. In this case, the function decays to nothing. Due to the bessel function, we have no $\lambda=0$ option. However, something interesting happens when $b^2 = \lambda$. Everything goes to zero, always, in that case, as sinh itself is always zero!

Basically, we have heat that dissipates here. Sinh may be a function that grows over time, but it's outcompeted by the exponential. (Proved by looking at the function in pure exponential form, the radical CANNOT be greater than b). 

But yes the behavior is exponential decay. Which is all we should see unless something really strange is going on.
