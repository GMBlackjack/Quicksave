# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Okay so we can note right away that multiplying by H leaves the last last two terms completley unchanged, as we can just say $H\beta = \sigma$ and similarly for q and be done with it. 

So what remains to be worked out is the first term, that is, the p-term. Let's just apply the product rule.

$$ \frac{d}{dx} (p \frac{d}{dx} \phi) = p \frac{d^2}{dx^2} \phi + \frac{d}{dx} p \frac{d}{dx}\phi $$

Anyway, we need to relate this with H times what we start with, so...

$$ p \frac{d^2}{dx^2} \phi + \frac{d}{dx} p \frac{d}{dx}\phi  = H \frac{d^2}{dx^2} \phi + H \alpha \frac{d}{dx}\phi $$

This would imply that H=p and $H\alpha = p'$. Since, technically speaking, we are free to specify p, all we need to find is a funciton such that $H\alpha = H'$ for any possible $\alpha$. The trivial answer is zero so that's pointless. However, we do know of funcitons that produce constants when differentiated:

$$ (e^{ax})' = (ax)'e^{ax} = ae^{ax} $$

We can do this with funcitons as well. Specifically, if we decalre $A(x) = \int \alpha(x) dx$, that is, A is the indefinite integral of $\alpha$ with respect to x, we can do the following:

$$ (e^{A(x)})' = e^{A(x)} \alpha(x) $$

And so...

$$p(x) = H(x) = e^{A(x)}$$
$$\sigma(x) = H(x)\beta(x) = e^{A(x)}\beta(x)$$
$$q(x) = H(x)\gamma(x) = e^{A(x)}\gamma(x)$$

We will note that technically A(x) can have an integration constant attached to it, but this is entirely arbitrary. We set it to zero since that is usually what's going to be the simplest to work with.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

Easy enough. First, divide by $x^2$ to get the form given at the start of **Problem 1**

$$ \frac{d^2\phi}{dx^2} + \frac{1}{x} \frac{d\phi}{dx} + \frac{\lambda \phi}{x^2} = 0 $$

Now, this means our $\alpha(x) = \frac{1}{x}$. The indefinite integral is then ln|x|.

Now, curiously, $H = e^{A(x)} = e^{ln|x|}$ is just |x|, so this simplifies things considerably already. Note that since we have an equality that equals 0 the actual influence of the multiplicaiton on the equality is nothing. 

$$ \frac{d}{dx} \left[ |x| \frac{d\phi}{dx} \right] + \frac{\lambda \phi |x|}{x^2} = 0 $$

or

$$ \frac{d}{dx} \left[ |x| \frac{d\phi}{dx} \right] + \frac{\lambda \phi}{|x|} = 0 $$

There is admittedly one problem here: the original equation was defined at x=0, but this last step is not, we have a division by the absolute value of |x|. Well, let's consider what would happen in the original equation if x=0. The only solutions that possibly work are $\lambda=0$, in which case $\phi$ is completley free to be defined as whatever, or $\phi=0$, in which case we have a trivial solution only. So even though the original equation was defined at x=0 it had no meaningful solution at that point, so we won't consider it as part of this solution. (Though we will return to this in part c), it appears). 

Also of note, our boudnary conditions only include 1 to b. Assuming b isn't negative, this excludes x=0 entirely. (And, in fact, would remove the absolute value sign as well.) 

The full Rayleigh quotien itself is kind of a complicated mess.

$$ \lambda = \frac{-p \phi \frac{d\phi}{dx}|_a^b + \int_a^b p\left( \frac{d\phi}{dx} \right)^2 - q\phi^2 dx}{\int_a^b \phi^2 \sigma dx} $$

However, 5.6.4 gives us the conditions for $\lambda \geq 0$ in a simpler light:

$$ -p \phi \frac{d\phi}{dx}|_a^b \geq 0 $$
$$ q \leq 0 $$

The second criteria is trivial as our q=0, for it does not exist. In the other case, however, we have p = |x| which on our bounds 1 to b is just x and thus doesn't change the sign obviously. 

$\phi$ itself, however, is 0 at *both* boundaries, meaning that the first criteria is met as everything is zero. Thus, $\lambda \geq 0$. So we are technically done. However, we are curious as to what exactly these conditions do to the Rayleigh quotient, so we spell it out clearly:

$$ \lambda = \frac{(positive) + \int_a^b p\left( \frac{d\phi}{dx} \right)^2 dx}{\int_a^b \phi^2 \sigma dx} $$

We note that p itself is positive (x) and that an integral of the remaining square must be positive as well, so there's no way for there to be any negative results. Mission accomplished. It does make you qonder why $p\geq 0$ isn't a requirement, since what remains could be negative without it, but in this case it's posisive so we need not be concerned. 

Of course, what of $\lambda = 0$...? This is dealt with later. 

So, we can tell rahter easily that $\lambda = 0$ nos not a valid result by examining the Rayleigh quotient. The "(positive)" in our specific case is zero, so we ignore it.

$$ \lambda = \frac{\int_a^b p\left( \frac{d\phi}{dx} \right)^2 dx}{\int_a^b \phi^2 \sigma dx} $$

Let's substitute in the exact values we know to get a celarer picture.

$$ \lambda = \frac{\int_1^b |x|\left( \frac{d\phi}{dx} \right)^2 dx}{\int_1^b \phi^2 /|x| dx} $$

Now, the absolute value of x is positive, the square of any function is always positive, so no matter what (unless $\phi$ is the trivial solutio nwe never care about) the result of these two integrals has to also be positive and finite. 

However, we are asked specifically to solve for the eigenvalues and eigenfunctions, so simply "outlining" the answer will not cut it, we need to be able to write down results. Specifically for

$$ \frac{d}{dx} \left[ |x| \frac{d\phi}{dx} \right] + \frac{\lambda \phi}{|x|} = 0 $$

We can expand this to a full differential equation rahter easily. Since we are in positive domain only due to boudnary conditions, the differential equation becomes:

$$ \phi'' + \frac{1}{x} \phi' + \frac{\lambda}{x^2}\phi = 0$$

Now... we need to solve this one. If we use the $x^r$ method, we can arrive at a result of $\phi = Ax^{i\sqrt\lambda}+Bx^{-i\sqrt\lambda}$. We know from previous work that $\lambda$ is positive. The problem is that working with this imaginary result is rather annoying. We can be pretty sure we're looking for an oscillatory answer, though, since we have boundary conditions of 0 at x=1 and x=b. So let's try to find a solution of the form $cos(X)$ where X is some unknown funciton of x. (Reasonably we expect since exponentials raised to imaginary numbers can be converted into trig funcitons, the same is true here, but we're not exactly confident in using that on any base other than e.) 

Assuming cos(X) we get the differential equation to become:

$$ -(X')^2cos(X) + X''sin(X)  + \frac{1}{x} X' sin(X) + \frac{\lambda}{x^2} cos(X) = 0 $$

The sine and cosine terms need to cancel. Which means $(X')^2 = \frac{\lambda}{x^2}$ This is an easy differential equation to solve: $X = \pm \sqrt{\lambda} ln(x)$. This also satisfies  $X'' = -X'/x$. 

So $cos(\sqrt{\lambda} lnx)$ is a solution. By a similar argument we see $sin(\sqrt{\lambda} lnx)$ is also a solution. Which means our total solution is:

$$ \phi(x) = Acos(\sqrt{\lambda} lnx) + Bsin(\sqrt{\lambda} lnx) $$

The sign of the component inside the trig functions is irrelevant, as that can be folded into A and B. 

Without initial conditions we cannot determine A and B, and therefore these are the eigenfunctions. However, we CAN determine the eigenvalues by the boundary conditions.

$$ \phi(1) = 0, \phi(b) = 0 $$

From this we actually put a restriction on the eigenfunctions. ln(1) is 0, but the cosine of zero is not zero! Thus, the cosine term must vanish, that is, every A=0.

$$ \phi(x) = Bsin(\sqrt{\lambda} lnx) $$

This matches the $\phi(1)$ condition. However, the $\phi(b)$ condition is where things get interesting. We want to force the interior of the sine function to be a multiple of pi at ln(b). Therefore...

$$ \lambda = \left(\frac{n\pi}{ln(b)}\right)^2 $$

Since we showed that zero is not an option, this starts at n=1. Making the smallest eigenvalue...

$$ \lambda_1 = \left(\frac{\pi}{ln(b)}\right)^2 $$

We will say that the sine of the natural log is a rather unusual funciton though, and one that only exists on the positive half of the domain, but that's all we care about so it's all good. 

The weight is asserted to be $\sigma$. In our case, we have $\sigma = 1/|x|$ or just 1/x since we're on the positive side.  
In order to demonstrate, we'd need to take the following integral and show it was zero:

$$ \int_1^b B_n sin\frac{n\pi lnx}{ln(b)}  B_m sin\frac{m\pi lnx}{ln(b)}\frac{1}{x} dx$$

The constant coefficients are irrelevant. 

$$ \int_1^b sin\frac{n\pi lnx}{ln(b)} sin\frac{m\pi lnx}{ln(b)}\frac{1}{x} dx$$

This looks like a case where we might be able to do u substitution. Let's say we let u = lnx, then du = 1/x dx. In which case, our bounds go from [1,b] to [0,ln(b)]. 

$$ \int_0^{ln(b)} sin\frac{n\pi u}{ln(b)} sin\frac{m\pi u}{ln(b)}du$$

And this... is exactly the same integral as from 0 to L for sine functions. Which we already know are orthogonal. Thus, we know the integral is zero, and we have shown the orthogonality property. 

This almost seems trivial. Take one of the equations:

$$ B_n sin\frac{n\pi lnx}{ln(b)} $$

lnx takes values ranging from 0 to ln(b), which is to say the argument of the sine function takes arguments from 0 to $n\pi$. Sine is known to have a zero at every single integer value of $\pi$. For n=1, the only zeroes are at the ends, adn thus do not qualify for our results, thus we have n-1=0 zeroes for n=1. 

For n=2, the sine function takes every value from 0 to $2\pi$. There will be a zero at $\pi$, but no others, so we have n-1=1 for n=2, as expected. 

In complete generality, pick an n. The sine funciton will have zeroes at 0 and $n\pi$, but we don't count those because they are on the ends. Between these two ends, however, every *other* integer value of $\pi$ will be hit, making a full list of zeroes $\pi, 2\pi, 3\pi, ... , (n-2)\pi, (n-1)\pi$ which is clearly a total of n-1. 

Now these zeroes are not evenly *spaced* since they depend on ln(x), but ln(x) is continuous from 0 to ln(b) so every point  *must* be hit along the way by way of continuity.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

First of all, we need to actually perform separation of variables. Assume u = TX. 

$$ c\rho (TX)_t = (K_0(TX_x))_x + \alpha XT $$
$$ \rightarrow Xc\rho T_t = T(K_0(X_x))_x + \alpha XT $$
$$ \rightarrow \frac{1}{T}c\rho T_t = \frac{1}{X}(K_0(X_x))_x + \alpha $$

Which we then turn into the separated equations

$$ c\rho T_t + \lambda T = 0 $$
$$ (K_0 X_x)_x + X\alpha + X\lambda = 0 $$

Now this gives us a $p=K_0$, $\sigma = 1$, and $q = \alpha$. 

As previously mentioned there are two conditions on the positive eigenvalues:

$$ -p \phi \frac{d\phi}{dx}|_a^b \geq 0 $$
$$ q \leq 0 $$

Note now the second one is only satisfied if $\alpha \leq 0$, which is partially what we wanted. (Strictly speaking, the two terms are not identical, as one allows $\alpha = 0$ and the other does not, but there does not appear to be a distinction here. Furthermore, the problem only asks us to show that the eigenvalues are positive IF $\alpha < 0$, **not** if and only if.) 

However, this still leaves the first condition. $\phi$ in our case is subject to boundary conditions at both ends, 0 and L, and at both of these places it is zero. Thus, the first condition is met because no matter what p is, $\phi$ will be zero, thus it vanishes, and the Rayleigh requirements are met for $\lambda \geq 0$. 

However, we are asked to specifically show that they are *negative*. So, with what remains, can we prove that the result MUST be positive and CANNOT be zero? With both conditions met, the remaining Rayleigh quotient is:

$$ \lambda = \frac{ \int_a^b p\left( \frac{d\phi}{dx} \right)^2 dx}{\int_a^b \phi^2 \sigma dx} $$



With our known values it becomes...

$$ \lambda = \frac{ \int_0^L K_0 \left( \frac{d\phi}{dx} \right)^2 dx}{\int_0^L \phi^2 dx} $$

Note that $K_0$ is a *function,* not a constant, so we can't just pull it out and evaluate. In fact, it seems as if we can prove that $K_0$ can be set to force the fraction to any value, since we know nothing about u--even making the integral negative, if $K_0$ was set to always have the opposite sign of $\phi_x^2$. 

From the textbook, we are apparently supposed to assume that $K_0$ is a **thermal coefficient** and is thus always positive. If this is the case and we do assume that it is not just some genreal function, then yes clearly $\lambda \geq 0$ still. 

Now, because $K_0$ has been assumed to be positive, which means, *not zero*, it means it cannot make the integral of $\phi_x^2$ zero either, as the square of anything has to always be positive. Thus, the eigenvalue **cannot be zero**. 

We're contnuing to assume $\alpha < 0$ here, so we know the eigenvalues will be positive. As for what exactly they are, well, that requires solving the differnetial equation.

$$ (K_0 X_x)_x + X\alpha + X\lambda = 0 $$

Which since we don't know $K_0$ we can't exactly do in any closed form. However, from the start of the problem we are told to assume we know all the eigenfunctions! Which technically means that the expression in the previous part is sufficient: If we knew every u and $K_0$ we could find every $\lambda$.

$$ \lambda_n = \frac{ \int_0^L K_0 \left( \frac{d\phi_n}{dx} \right)^2 dx}{\int_0^L \phi_n^2 dx} $$

THe INITIAL value problem, however, will ask us to find the coefficients... aha, we see what's being asked. For that, let's examine the situation for the full solution. First, we have to examine the temporal side.

$$ c\rho T_t + \lambda T = 0 $$

Which has the somewhat evident solution of 

$$ T = Ae^{-\frac{\lambda}{c\rho} t} $$

Now if we know the eigenfunctions, we can call the whole solution:

$$ \sum_{n=1}^\infty A_n \phi_n e^{-\frac{\lambda}{c\rho} t} $$

The initial value problem dictates that, at t=0...

$$ f(x) = \sum_{n=1}^\infty A_n \phi_n $$

To solve for the coefficient we use the orthogonality relation. This requires $\sigma$, which is just 1 as we found out earlier so the eigenfunctions must be directly orthogonal. Multiplying by $\phi_m$ and integrating, as usual, we get...

$$ \int_0^L f(x) \phi_m dx = A_m  \int_0^L \phi_m^2 dx $$

Which is sufficient for the solution, since we are told to assume we know the eigenfunctions $\phi_n$.

$$ \frac{\int_0^L f(x) \phi_m dx}{\int_0^L \phi_m^2 dx} = A_m $$ 

Pretty clearly as t approaches infinity, the full solution exponentially decays into nothing. Not surprising in the slightest, pretty standard for this sort of problem, really, and will eventually overrule anything X has to offer. 

We will note that T contians $\lambda$, however since we know $\lambda$ is positive the overall behavior at large t does not change, just the rate at which it does change. We know that there will be a smallest eigenvalue, and so that term will dominate as it decays the slowest, and that every larger eigenvalue drops off much faster. So if we're looking at behavior as t approaches infinity, it is best to use $\lambda_1$ with n=1 (assuming we set it up so n=1 was the smallest. It would make sense if it was, but there are unusual cases that exist.)

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

This can roughly be rephrased as $uv_{xxxx} - vu_{xxxx}$. We seek to showt aht this is an exact differential. The question is, how exactly do we define an exact differential? Well, the way it was done in teh book was to show that it was equal to the first derivative of something else. 

So, we scribbled down a lot of derivatives and eventually found one. Bear with us, this is a bit counterintuitive. We went seraching for something of the form:

$$ \frac{d}{dx}(f) = uv_{xxxx} - vu_{xxxx} $$

We reasoned that the only way this could be the case is if two terms exted in f that would produce the results as part of the product rule. However, there would have to be something additional to cancel out the other terms.

$$ \frac{d}{dx}(uv_{xxx} - vu_{xxx} + A) = uv_{xxxx} - vu_{xxxx} $$

Now A would end up part of an equation all its own.

$$ \frac{d}{dx} A = - u_x v_{xxx} + v_x u_{xxx} $$

Now this one can actually be solved directly since *we are close to having the number of derivatives match*. Behold:

$$ \frac{d}{dx} (-u_xv_{xx} + v_xu_{xx}) = -u_xv_{xxx} - u_{xx}v_{xx} + v_xu_{xxx} + v_{xx}u_{xx} = - u_x v_{xxx} + v_x u_{xxx} $$

Which we can sibstitute back into our relaiton to show

$$ \frac{d}{dx}(uv_{xxx} - vu_{xxx} -u_xv_{xx} + v_xu_{xx}) = uv_{xxxx} - vu_{xxxx} $$

And we have demonstrated that it is an exact differential! 

Since we have the differnetial form, this is somewhat trivial:

$$ \int_0^1 uv_{xxxx} - vu_{xxxx} dx = [uv_{xxx} - vu_{xxx} -u_xv_{xx} + v_xu_{xx}]_0^1$$

$$ [uv_{xxx} - vu_{xxx} -u_xv_{xx} + v_xu_{xx}]_0^1 $$

This is our result from part b). The first boundary condition makes the funcitons u(0) v(0) u(1) and v(1) all equal zero, so the first two terms never ammount to anything, thereby removing our need to know anything about hte third derivatives. 

We can only say that u'(0) and v'(0) equal zero, we don't know about u'(1). But for every u'(1) there is a v''(1) and vice versa, and both u''(1) and v''(1) = 0, so the last two terms always go to zero as well. Therefore...

$$ [uv_{xxx} - vu_{xxx} -u_xv_{xx} + v_xu_{xx}]_0^1 = 0$$

Before we move on, let's consider what we actually have here. We have the equivalent of Lagrange's Identity for this operator L that's the fourth derivative, and from that we have the Green's formula equivalent. in this latest part where we showed the integral was zero, we proved the self-adjoint property--this is a property of L, not of the functions u and v, u and v are just some eigenfunctions of the system.

This means that even though this isn't an Strum-Liouville situation, the orthogonality relations done for S-L should still apply here as we are still self-adjoint. 

We can simplify this expression into operator notation.

$$ L(\phi) + \lambda e^x \phi = 0 $$

This looks remarkably similar to 5.5.10, giving us a clue. Let's declare two different eigenfuncitons. We know we have at least two since we were provided u and v before. 

$$ L(\phi_n) + \lambda_n e^x \phi_n = 0 $$
$$ L(\phi_m) + \lambda_m e^x \phi_m = 0 $$

Let's explicitl6y put this into our earlier integral relation. 

$$ \int_0^1 \phi_nL(\phi_m)- \phi_mL(\phi_n) dx = [\phi_n \phi_{mxxx} - \phi_m\phi_{nxxx} -\phi_{nx}\phi_{mxx} + \phi_{mx}\phi_{nxx}]_0^1$$

Now since we don't want to write the right side a million times we just set it to zero since we know it has to be.

$$ \Rightarrow \int_0^1 \phi_nL(\phi_m)- \phi_mL(\phi_n) dx = 0$$

We can replace the things L is operating on with our differential equation relation.

$$ \Rightarrow \int_0^1 - \phi_n\lambda_m e^x \phi_m+ \phi_m\lambda_n e^x \phi_n dx = 0$$

We can re-arrange this into an orthogonality relation.

$$ \Rightarrow (\lambda_n - \lambda_m)\int_0^1  \phi_n\phi_me^x dx = 0$$

This proves the orthogonality of the (still unknown) eigenfunctions, and gives the weighting function for free: $e^x$. Keep in mind this is not universal for all L, just self-adjoint L. WHich is what the first three parts of this problem were about showing. 

Seems like we want the Rayleigh Quotient... except we aren't in a S-L differential equation, which complciates matters, we can't just assume the derivation will be the same. However, we can follow similar procedures to the original derivation and see where we end up.

$$ L(\phi) + \lambda e^x \phi = 0 $$

Multiply through by $\phi$

$$ \Rightarrow \phi L(\phi) + \lambda e^x \phi^2 = 0 $$

Now integrate.

$$ \Rightarrow \int_0^1 \phi L(\phi) dx + \lambda \int_0^1 \phi^2 e^x dx = 0 $$

Which *does* give us an expression.

$$ \Rightarrow \lambda = \frac{- \int_0^1 \phi L(\phi) dx}{\int_0^1 \phi^2 e^x dx} $$

Notably the denominator is the same as the standard Rayeligh quotient. However, what can the numerator tell us? 

Well, what we do is integrate by parts not once, but twice. First time...

$$ \Rightarrow \lambda = \frac{- \phi \phi_{xxx} |_0^1 + \int_0^1 \phi_x \phi_{xxx} dx}{\int_0^1 \phi^2 e^x dx} $$

That evaluation goes to zero. 

$$ \Rightarrow \lambda = \frac{ \int_0^1 \phi_x \phi_{xxx} dx}{\int_0^1 \phi^2 e^x dx} $$

Integrate by parts agian.

$$ \Rightarrow \lambda = \frac{  \phi_x \phi_{xx} |_0^1- \int_0^1 \phi_{xx}^2 dx}{\int_0^1 \phi^2 e^x dx} $$

The evaluation goes to zero once more, leaving us with

$$ \Rightarrow \lambda = \frac{ - \int_0^1 \phi_{xx}^2 dx}{\int_0^1 \phi^2 e^x dx} $$

The integral in the numerator will always be positive due to the squaring, and so the minus sign out front ensures that the eigenvalue will be negative. But can it be zero? No! Only if $\phi$ itself were zero would there be a zero eigenvalue, and we don't care about *that* situation! (one could also suggest the second derivative being zero would work, but we know that isn't the case since the differential equation works with fourth derivatives.)

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$



Bring in the Rayleigh coefficient again. To actually determine this, we need to determine how this correlates to the Sturm-Liouville form. It actually is quite easy: p=1, $\sigma = 1$, and $q = -x^2$

The conditions we have to specify are, as usual:

$$ -p \phi \frac{d\phi}{dx}|_a^b \geq 0 $$
$$ q \leq 0 $$

q is clearly less than zero, being negative. The first condition is also met, since our bounds a to b are 0 to 1 in this problem, which means that the derivative portion equals zero at both bounds and sets the entire first condition to zero. We also check that p is always positive, it is equal to 1. 

However, this only confirms positive OR zero eigenvalues. We need to prove that zero is NOT a solution! To do that, we take what remains of the Rayleigh quotient:

$$ \lambda = \frac{ \int_0^1 p\left( \frac{d\phi}{dx} \right)^2 - q\phi^2 dx}{\int_a^b \phi^2 \sigma dx}  = \frac{ \int_0^1 \left( \frac{du}{dx} \right)^2 + x^2u^2 dx}{\int_0^1 u^2 dx} $$

Note that we have an extra term that usually isn't there, the $x^2$. However, *every term in the integral is squared*, which means the function under the integral is always positive and has no chance to be zero except at x=0, and for a definite integral this is not sufficient to make the integration go to zero. With both numerator and denominator confirmed positive, we can rule out the zero case and say $\lambda > 0$.

The trial function method relies on u being replaced with some $u_T$ via 5.6.6

$$ \lambda_1 \leq RQ[u_T] = \frac{-pu_T \frac{du_T}{dx}|_a^b + \int_a^b [p(\frac{du_T}{dx})^2 + qu_T^2]dx}{\int_a^b u_T^2 \sigma dx} $$

Now our p, q, and $\sigma$ are the same as before so this quickly simplifies.

$$ \lambda_1 \leq  \frac{ \int_0^1 [(\frac{du_T}{dx})^2 + x^2u_T^2]dx}{\int_0^1 u_T^2 dx} $$

Now before we continue we need to carefully specify our function and make sure it matches boundary conditions. First of all, let's look for the roots. $(x-x^2)^2+\alpha = 0$. We want to find none between 0 and 1. 

$$ x^4 - 2x^3 + x^2 + \alpha = 0 $$

Since we don't know $\alpha$, why not examine the equation without it?

$$ x^4 - 2x^3 + x^2 = x^2(x^2 - 2x + 1) = x^2(x-1)^2 $$

This means we have four roots, two at 0 and two at 1. This is fine, there are no middle zeroes. Raising by a positive $\alpha$ wouldn't change this, either, as this is an even power system, so all the values are positive no matter what. However, if we were to *subtract*, there are cases where there would be other zeroes, however after enough of a movement (currently unknown) the only zeroes would be the crossings of the x axis as x goes to both infinities, which will be outside of our bounds. So there are *lots* of options for $\alpha$. (We believe 0 will be the sharpest, though). Just note that positive values work, 0 works, and some negative (but not all negative!) values work. 

One thing we also need is the derivative to be zero at x=0 or x=1. This is obviously met at x=0, but 1?

$$ 4x^3 - 6x^2 + 2x = 4-6+2 = 0 $$

Confirmed. The presence of $\alpha$ has no influence whatosever on matching the boundary conditions. 

We do feel like checking where the negative $\alpha$ values won't work. We can find them by finding the maxima that occurs in the middle of the graph--we have an even power that stays positive at all times, there has to be a hump in there, it'll be the maximum. Also since this is clearly a symmetric situation it'll be right in the middle at x=0.5. The derivative is, in fact, zero here. 

The actual value of the function is... 1/16. Plus alpha, of course. So anything more negative than $\alpha = -1/16$ is valid. So our prohibited $\alpha$ range is officially [-1/16,0), with the specific note that 0 is allowed so it is an open bound.



Anyway, retu4rning to our inequality, we can make the substitution. 

$$ \lambda_1 \leq  \frac{ \int_0^1 [(4x^3 - 6x^2 + 2x)^2 + x^2(x^4 - 2x^3 + x^2 + \alpha)^2]dx}{\int_0^1 (x^4 - 2x^3 + x^2 + \alpha)^2 dx} $$

After some algebra, we arrive at:

$$ \lambda_1 \leq  \frac{ \int_0^1 x^{10} - 4x^9 + 6x^8 -4x^7 + (17+2\alpha)x^6 + (-48 - 4\alpha)x^5 + (52+2\alpha) x^4 - 24x^3 + (4+\alpha^2) x^2 dx}{\int_0^1 x^8 - 4x^7 + 6x^6 - 4x^5 + (1+2\alpha)x^4 -4\alpha x^3 + 2\alpha x^2 + \alpha^2 dx} $$

Now, obviously, as polynomial definite integrals, the integral at zero goes to zero no matter what. So all that really matters is the integral at 1. Since the power rule is very obvious and we don't want to write all that out again, we will do the integration and substitution step all at once since when x=1 the powers up and vanish, simplifying everything consierably. 

$$ \lambda_1 \leq \frac{ \frac{1}{11} - \frac{4}{10} + \frac{6}{9} - \frac{4}{8} + \frac{17+2\alpha}{7} - \frac{48+4\alpha}{6} + \frac{52+2\alpha}{5} - \frac{24}{4} + \frac{4+\alpha^2}{3}}{ \frac{1}{9} - \frac{4}{8} + \frac{6}{7} - \frac{4}{6} + \frac{1+2\alpha}{5} - \frac{4\alpha}{4} +\frac{2\alpha}{3} + \alpha^2 } $$

Now normally we'd just add everything together in a big number and get the result, but we wish to MINIMIZE the result by controlling $\alpha$. So let's look at how much $\alpha$ contributes on top and bottom: does it make things bigger? Smaller? Otherwise?

$$ \frac{... + \frac{2\alpha}{7} - \frac{4\alpha}{6} + \frac{2\alpha}{5} + \frac{\alpha^2}{3} }{... + \frac{2\alpha}{5} - \frac{4\alpha}{4} +\frac{2\alpha}{3} + \alpha^2 } $$

Clearly the magnitdue of $\alpha$'s influence is lesser in the numerator than the denominator, given that it is divided by larger numbers. However, there is no clear indicator if its influence is positive or negative, and how it depends on $\alpha$'s sign. 

Since we are going for an estimation we opt to stop going for an exact answer and reduce fracitons to decimal form. 

$$ \lambda_1 \leq \frac{ 0.01948 + 0.01905\alpha + 0.33333\alpha^2}{0.00159 + 0.06667 \alpha + \alpha^2 } $$

This is a little easier to understand.

First and foremost, here's the graph, with -1/16 marked.



This is a bit opposite to what we expected--we thought either 0 or -1/16 would be the best. However, this appears to not be the case, becuase the "best" is the harshest limit on what $\lambda$ can be, and that is found at the limits as $\alpha$ goes to infinity. And even though we took all this time to add all the fracitons together, THAT limit is known exactly. No matter if one goes positive or negative, the absolute minimum of this function is...

$$ \lambda_1 \leq \frac{1}{3} $$

The answer is neat, but we were still somewhat unconvinced that taking a limit was the solution. But even having a computer do all the integration directly, each step was confirmed, and the end result produced the exact same values for every $\alpha$ that we ended up with. So, apparently, increasing $\alpha$ without bound minimizes $\lambda$ more and more. One wonders why this would be the case and does not find any real solutions. The result for $\lambda$ itself, though, seems reasonable as a sensible upper bound. 

We note that even in the "dissalowed" area between 0 and -1/16, the estimation for $\lambda$ is still positive. It's just really really high. 0 is in fact a better estimator than -1/16.
