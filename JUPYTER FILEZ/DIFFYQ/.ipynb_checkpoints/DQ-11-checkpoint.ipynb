# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

The eigenvalues can be determined via 5.9.11 in an approximation:

$$ \lambda \approx \left[ \frac{n\pi}{\int_0^L \sqrt\frac{\sigma}{p} dx}\right]^2 $$

Our L is 1 in this case from the boundary conditions. p is $(1+x)^2$. $\sigma$  is $1+x$. With this, we end up with:

$$ \lambda \approx \left[ \frac{n\pi}{\int_0^1 \sqrt\frac{1}{1+x} dx}\right]^2 $$
$$ = \left[ \frac{n\pi}{\left[ 2\sqrt{x+1} \right]_0^1}\right]^2 $$
$$ = \left[ \frac{n\pi}{2\sqrt{2}-2}\right]^2 $$
$$ = \frac{n^2\pi^2}{4(\sqrt{2}-1)^2}$$

This is the approximation for large $\lambda$. Keep in mindt his was derived for $\phi(0)=0$, which it does in our case.

There also exists an approximation formula for the functions, 5.9.8.

$$ \phi(x) \approx (\sigma p)^{-1/4} exp\left[ \pm \sqrt\lambda \int^x \sqrt\frac{\sigma}{p} dx_0 \right] $$

We determine the function itself by substituting and applying $\phi(0) = 0$

$$ \phi(x) \approx ((1+x)^3)^{-1/4} exp\left[ \pm \frac{n\pi}{2\sqrt{2}-2} \int_0^x \sqrt\frac{1}{1+x} dx_0 \right] $$

$$ = (1+x)^{-3/4} exp\left[ \pm \frac{n\pi}{2\sqrt{2}-2} (2\sqrt{x+1} - 2) \right] $$
$$ = (1+x)^{-3/4} exp\left[ \pm n\pi\frac{2\sqrt{x+1} - 2}{2\sqrt{2}-2} \right] $$

Which are some... interesting functions to say the least. With our boundary conditions it's required that we turn this into a sine function, which is easy enough to do. 

$$ = (1+x)^{-3/4} sin\left[ n\pi\frac{2\sqrt{x+1} - 2}{2\sqrt{2}-2} \right] $$

Which is to say plot n=1, 2, 3, 4, which are supposedly the "least accurate" of the eigenvalues. Strangely there's no request here about what exactly to check this against, so we shall refrain from doing so.

Code:

```
x=0:0.001:400;
y = (1+x).^(-3/4).*sin(pi * (2.*(sqrt(x+1))-2)/(2.*sqrt(2)-2));
y2 = (1+x).^(-3/4).*sin(2.*pi * (2.*(sqrt(x+1))-2)/(2.*sqrt(2)-2));
y3 = (1+x).^(-3/4).*sin(3.*pi * (2.*(sqrt(x+1))-2)/(2.*sqrt(2)-2));
y4 = (1+x).^(-3/4).*sin(4.*pi * (2.*(sqrt(x+1))-2)/(2.*sqrt(2)-2));
plot(x,y);
xlim([0,1]);
ylim([-2,2]);
xlabel('x'); 
ylabel('phi');
hold
plot(x,y2);
plot(x,y3);
plot(x,y4);
```

Taking this one step at a time, first we need to find the Fourier sine series. This is not a funciton we've done before, so lets grab the series:

$$ f(x) \text{~} \sum_{n=1}^\infty B_n sin\frac{n\pi x}{L} $$
$$ B_n = \frac 2L \int^L_0 f(x) sin\frac{n\pi x}{L} dx $$

Since we need exact values in our answer, the results of $B_n$ are quite necessary, so let's evaluate them.

$$ B_n = 2 \left( \int^{1/2}_0 x sin(n\pi x) dx + \int^{1}_{1/2} sin(n\pi x) dx - \int^{1}_{1/2} x sin(n\pi x) dx \right)$$

$$ = 2 \left( \left[ \frac{sin(n\pi x)}{n^2\pi^2} - x\frac{cos(n\pi x)}{n \pi} \right]^{1/2}_0 -\left[ \frac{cos(n\pi x)}{n\pi}\right]^{1}_{1/2} - \left[ \frac{sin(n\pi x)}{n^2\pi^2} - x\frac{cos(n\pi x)}{n \pi} \right]^{1}_{1/2} \right)$$

$$ = 2 \left( \frac{sin(n\pi/2)}{n^2\pi^2} - \frac{cos(n\pi/2)}{2 n \pi} - \frac{cos(n\pi)}{n\pi} + \frac{cos(n\pi /2)}{n\pi} + \frac{cos(n\pi)}{n \pi} + \frac{sin(n\pi/2)}{n^2\pi^2} - \frac{cos(n\pi/2)}{2n \pi} \right)$$

First of all, we find the Fourier Sine Series already done in 3.3.11 and 3.3.12.

$$ x = \sum_{n=1}^\infty \frac{2L}{n\pi}(-1)^{n+1} sin\frac{n\pi x}{L} $$

So, now we ask, waht is Parseval's Identity? Well, it is 5.10.14...

$$ \int_a^b f^2 \sigma = \sum_{n=1}^\infty \alpha_n^2 \int_a^b \phi_n^2 \sigma dx $$

Well we can apply this to the sine series. 

$$ \int_0^L x^2 = \sum_{n=1}^\infty \frac{4L^2}{n^2\pi^2} \int_0^L sin^2 \frac{n\pi x}{L}dx $$

Which rather simply evaluates...

$$ \frac{L^3}{3} = \frac{4L^2}{\pi^2} \frac{L}{2} \sum_{n=1}^\infty \frac{1}{n^2} $$
$$ \Rightarrow \frac{\pi^2}{6} = \sum_{n=1}^\infty \frac{1}{n^2} $$

And that is our answer. 

Don't remember evaluating that series before so we'll have to get it manually. Shouldn't be too hard, though, $x^2$ is already even and has well known results.

$$ f(x) = \sum_n A_n cos\frac{n\pi x}{L} $$
$$ A_0 = \frac{1}{L} \int_0^L f(x) dx $$
$$ A_n = \frac{2}{L} \int_0^L f(x) cos\frac{n\pi x}{L} dx$$

$$ A_0 = \frac{L^2}{3} $$

$$ A_n = \frac2L \frac{2L^3}{n^2 \pi^2} cos(n\pi) = \frac{4L^2}{n^2 \pi^2} cos(n\pi) $$

Now, with Parseval's identity... note the n=0 case exists, but isn't being done ofer a cosine funciton, but a cosine(0) situation, which is just 1^2 which integrates to L. 

$$ \int_0^L x^4 = \frac{L^5}{9} +  \sum_{n=1}^\infty A_n^2 \int_0^L cos^2 \frac{n\pi x}{L}dx $$
$$ \Rightarrow \frac{4L^5}{45} = \sum_{n=1}^\infty A_n^2 \frac{L}{2} $$
$$ \Rightarrow \frac{8L^4}{45} = \sum_{n=1}\frac{16L^4}{n^4\pi^4}$$ 
$$ \Rightarrow \frac{\pi^4}{90} = \sum_{n=1}\frac{1}{n^4}$$ 

Which is exactly what it should be.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

So, let u = $\phi$h. Then we can show...

$$(\phi h)_{tt} = c^2 ((\phi h)_{xx} + (\phi h)_{yy})$$
$$\Rightarrow \phi h_{tt} = c^2 (h \phi_{xx} + h \phi_{yy})$$
$$\Rightarrow \frac1h h_{tt} = c^2 \frac1\phi (\phi_{xx} + \phi_{yy})$$

The variables have been separated into t and xy. Setting them to be equal to lambda, we clearly have...

$$ \frac1h h_{tt} = -\lambda $$
$$ -\lambda = c^2 \frac1\phi (\phi_{xx} + \phi_{yy})$$

At the moment we only care about the xy portion, so take that.

$$ c^2 \frac1\phi (\phi_{xx} + \phi_{yy}) + \lambda = 0$$

$$ \Rightarrow (\phi_{xx} + \phi_{yy}) + \lambda \frac{1}{c^2} \phi= 0$$

$$ \Rightarrow \nabla^2 \phi + \lambda \frac{1}{c^2} \phi= 0$$

Which makes $\sigma = 1/c^2$.

The fact that $\phi = 0$ on the boundary almost goes without saying, since the individual $\phi$ values still have to satisfy the differential equation for x and y. 

We note that the form of this equation is very similar to the form used to prove the orthtogonality for the Sturm-Liouville problem. 

$$ L(\phi) + \lambda \sigma(x,y) \phi = 0 $$

In fact the only real difference is that the function $\sigma$ is now in x and y instead of just x. To prove orthogonality, we'll want to explicitly label our eigenfunctions and eigenvalues.


$$ L(\phi_n) + \lambda_n \sigma(x,y) \phi_n = 0 $$
$$ L(\phi_m) + \lambda_m \sigma(x,y) \phi_m = 0 $$

However to get much further here we need to derive a version of Green's formula for the laplacian.

$$ \int (u\nabla^2v - v\nabla^2u) dxdy = ? $$

7.5.7 gives this to us.

$$ \int (u\nabla^2v - v\nabla^2u) dxdy = \oint (u\nabla v - v \nabla u) \cdot \hat n dx $$

Since our equations have to satisfy the boundary conditions, we can move forward. So now we set up:

$$ \int\int_R (\phi_m \lambda_n \sigma(x,y) \phi_n - \phi_n \lambda_m \sigma(x,y) \phi_m) dxdy$$
$$ = \int\int_R (-\phi_m L \phi_n + \phi_n L \phi_m) dxdy$$
$$ = \oint_R (-\phi_m \nabla \phi_n + \phi_n \nabla \phi_m) \cdot \hat n ds$$

We note that this must be zero since the funcitons $\phi$ are zero on the boundaries, and we are integrating ONLY over the boundary. This means that...

$$ 0 = \int\int_R (\phi_m \lambda_n \sigma(x,y) \phi_n - \phi_n \lambda_m \sigma(x,y) \phi_m) dxdy$$
$$ =  (\lambda_n - \lambda_m) \int\int_R (\phi_m \sigma(x,y) \phi_n) dxdy$$

Which is the definition of orthogonality with a "weight" $\sigma$. And so we are done. 

So, borrowing the method from Section 5.5, we note first assume that an eigenvalue can be complex.

$$ L(\phi) + \lambda \sigma(x,y) \phi = 0 $$

If imaginary numbers are allowed, the complex conjugate should also work.

$$ L(\phi)^* + \lambda^* \sigma(x,y) \phi^* = 0 $$

This is assuming $\sigma$ is wholly real, naturally. This is sensible since, realistically speaking, what would an imaginary relation between time and space on a vibrating membrane even look like? Unphysical! We note that L, being wholly real, only has the conjugate apply to the function inside.

$$ L(\phi^*) + \lambda^* \sigma(x,y) \phi^* = 0 $$

Naturally both $\phi$ and $\phi^*$ must satisfy the same boundary conditions. 

Due to orthogonality, the following must be true:

$$ (\lambda - \lambda^*) \int\int_R \phi \phi^* \sigma dxdy = 0 $$

But a basic property of complex conjugates is that $\phi \phi^* = |\phi|^2$ which, as a magnitude, is greater than zero and thus means the integral cannot be true! Ergo, complex eigenvalues are not possible. 

We can modify the 2D Rayleigh Quotient to our purposes here. 

$$ \nabla^2\phi + \lambda \sigma(x,y) \phi = 0 $$
$$ \Rightarrow \lambda = -\frac{\int\int_R \phi \nabla^2 \phi dxdy}{\int\int_R \phi^2 \sigma dxdy} $$

Notably this only changes the denominator, so 7.6.5 is only adjusted slightly.

$$ \lambda = \frac{-\oint \phi \nabla \phi \cdot \hat n ds + \int\int_R | \nabla \phi |^2 dxdy}{\int\int_R \phi^2 \sigma dxdy} $$

From before, we know any integral across the boundary has to be zero, due to the boundary conditions being held at zero.

$$ \lambda = \frac{\int\int_R | \nabla \phi |^2 dxdy}{\int\int_R \phi^2 \sigma dxdy} $$

The numerator, being the integral of a magnitude, must be positive (and not zero!). The denominator is a square function times another nonzero funciton ($\sigma$) so it too must be positive. Teherefore, $\lambda > 0$. And we are done.

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$

There are a lot of ways to write out the Gram-Smidt procedure, but the textbook states it like so on page 292:

$$ \psi_j = \phi_j - \sum_{i=1}^{j-1} \left( \frac{\int\int_R \phi_j\psi_i dxdy}{\int\int_R \psi_i^2 dxdy}\right) \psi_i $$

Notably this is two dimensional, when our result is purely one-dimensional, so we will have to adapt it. $\phi$ are the original "vectors", that is the linearly independent system we are provided, and $\psi$ are what we seek. Also note that we start at zero, not one. Adjusting to 1D on our provided interval changes this to:

$$ \psi_j = \phi_j - \sum_{i=0}^{j-1} \left( \frac{\int_{-1}^1 \phi_j\psi_i dx}{\int_{-1}^1 \psi_i^2 dx}\right) \psi_i $$


As anyone who's used the GS procedure will say, things can get really out of hand really quickly. So we will tackle it one term at a time. The first term is simple and is right out front withotu any issues. Just substitute j=4 and we get

$$ x^4 $$

The somewaht trivial "0" term is...

$$ - \frac12 \int_{-1}^1 x^4 dx = - 1/5  $$

The next term is 

$$ - \left( \frac{\int_{-1}^1 \phi_4\psi_1 dx}{\int_{-1}^1 \psi_1^2 dx}\right) \psi_1 = - \left( \frac{\int_{-1}^1 x^4 x dx}{\int_{-1}^1 x^2 dx}\right) x = 0 $$

Numerator was odd, this term goes away. makes sense, given the fact that the Legandre polynomials have a clear pattern of alternating terms like that. Next...

$$ - \left( \frac{\int_{-1}^1 \phi_4\psi_2 dx}{\int_{-1}^1 \psi_2^2 dx}\right) \psi_1 = - \left( \frac{\int_{-1}^1 x^4 (x^2-1/3) dx}{\int_{-1}^1 (x^2-1/3)^2 dx}\right) (x^2-1/3) = - \left( \frac{16/105}{8/45}\right) (x^2-1/3) = -6/7(x^2-1/3) $$

Contributes to a squared and a constant term, which is what we expect. 

We don't even need to write out number 3, as it will once again make the numerator odd and integrate to nothing. 



Now, combine everything together...

$$ x^4 -6x^2/7 + 6/21 - 1/5 = x^4 - 6x^2/7 + 3/35 $$

Which does in fact match the known $P_4$ of $35x^4 - 30x^2 + 3$, though when fully normalized the true one has a 1/8 out front, but those were left off the given polynomials so we shall not consider it here. 

Obviously start with $\psi_4$ and start subtracting off the other terms. With no $x^3$ term we ignore $\psi_3$ and move directly to $\psi_2$. We set it to get rid of the $x^2$ term...

$$ \psi_4 + (6/7) \psi_2 = x^4 + 3/35 - 2/7 = x^4 - 7/35 = x^4 - 1/5 $$

Which means our last term is just adding 1/5. So all together..

$$ x^4 = \psi_4 + (6/7) \psi_2 + (1/5) \psi_0 $$



Consider cosx and cos2x to be our "components". Let's just declare the first one to be the base "vector."

$$ \psi_1 = cosx + cos(2x) $$

Which could be effectively represented as (1,1). 

Now, use the GS procedure (as outlined in the text, not as used elsewhere) to get another one.

$$ \psi_2 = \phi_2 - \left( \frac{\int_{0}^\pi \phi_2\psi_1 dx}{\int_{0}^\pi \psi_1^2 dx}\right) \psi_1 $$

Fortunately, there is only one of the terms, so the sum has vanished. We can keep track of it all at once. Substitutions...

$$ = (3cosx - 4cos2x) - \left( \frac{\int_{0}^\pi (3cosx - 4cos2x)(cosx + cos2x)dx}{\int_{0}^\pi (cosx + cos2x)^2 dx}\right) (cosx + cos2x) $$

$$ = (3cosx - 4cos2x) - \left( \frac{\int_{0}^\pi (3cos^2x + 3cosxcos2x - 4cosxcos2x - 4cos^22x)dx}{\int_{0}^\pi (cos^2x +2cosxcos2x +cos^22x) dx}\right) (cosx + cos2x) $$

Several terms just integrate to zero, remove them.

$$ = (3cosx - 4cos2x) - \left( \frac{\int_{0}^\pi (3cos^2x - 4cos^22x)dx}{\int_{0}^\pi (cos^2x  +cos^22x) dx}\right) (cosx + cos2x) $$

And these square integrals not only ar eknown, we've been doing a *lot* of them lately. Furthermore, the presence of the "2" has no effect, since the component that surfvives is the x/2 component! All results are $\pi/2$ due to this.

$$ = (3cosx - 4cos2x) - \left( \frac{ (-\pi/2)}{\pi}\right) (cosx + cos2x) $$


$$ = (3cosx - 4cos2x) +  \frac{ 1}{2} (cosx + cos2x) $$

$$ = ((3.5)cosx - (3.5)cos2x) $$


Which we can write as (3.5,-3.5) in our system. Which is obviously orthogonal to (1,1) if we think of it as a vector space, so we consider this a success. Notably it is not normalized, but we do not need to do this. (That said, it would be trivial to normalize to (1,-1).)
