# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$



This is written out in full for the 2D form on page 81, naturally this must be very similar. If the Laplacian is satisfied in 3D, that is, $\nabla^2 u = 0$, then 

$$ 0 = \int\int\int \nabla^2 u d^3r  = \int\int\int \nabla \cdot (\nabla u) d^3r $$

The next step is a rather well-known integral identity that is stated in Zangwill's *Modern Electrodynamics* (without naming it, though we know it's the divergence theorem) as $\int_V \nabla \cdot \textbf{A} = \int_S \hat n \cdot \textbf{A}$. This means we can now directly state:

$$ 0 = \oint\oint \hat n \cdot (\nabla u)  dS = \oint\oint (\nabla u) \cdot \hat n dS   $$

S is specifically a closed surface, hence the closed integrals. Technically speaking the "closed bounds" should be a single circle around both integrals. Our LaTeX doesn't have that symbol enabled for some reason and we don't want to troubleshoot at the moment. 

As for the physical interpretation, this is actually a relatively well-known one. The amount of heat going into an area has to equal the amount going out, no matter what the area is, what it shape is, or anything. Laplace's equation demands that the solution be steady state, so naturally it demands that any "flow" always be exactly countered so the heat density doesn't actually change. Note that particles can still flow so long as flow is countered everywhere always.

In Electromagnetics we often think of this as there being "no net flux."

For illustrative purpsoes, we can think of two. One is just some object of constant temperature in deep space that isn't radiating, so it just maintains its temperature everywhere always. But we can think of another one to illustrate the curiosity of the "allowance of flow." Imagine a ring where heat can only flow clockwise. The heat flows out of every section in one direction, but since it is a ring, it always flows back in. Thus, even a flowing system can still astisfy Laplace's equation.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$



Since we are not provided an indefinite itnegral to work with, unlike the previous homework, we assume we have to figure it out ourselves. Fortunately, the trig function product-to-sum formulas are well known. We can immediately say:

$$ sin(n\pi x/L)sin(m\pi x/L) = \frac12 [cos((n-m)\pi x/L) - cos((n+m)\pi x/L)] $$

Of which the indefinite integral is

$$ \frac12 [\frac{L}{(n-m)\pi}sin((n-m)\pi x/L) - \frac{L}{(n+m)\pi}sin((n+m)\pi x/L)] $$

Now as we already know from **Problem 2** in the previous homework, this is only valid for $m \neq n$, there are other edge cases to deal with. But we can still evaluate this. At x=0, all of these terms go to zero. At x=L, all of the terms go to zero, since everything is taking the sine of a multiple of pi. So that's taken care of. 

When n=m=0, the original integral becomes flat zero. It can never equal anything else. 

When m=n, then we have something a little different, the integraal becomes that of a square sine function:

$$ \int_0^L sin^2(n\pi x/L) dx $$

Which is a known integral that evaluates to

$$ \left[ -\frac{L}{4 n\pi} sin(2 \frac{n\pi x}{L}) + \frac x2 \right]_0^L $$

Which evaluates to L/2. This means our total answer is:

0 when $n \neq m$, 0 when $n=m=0$, and $\frac L2$ when $n=m\neq0$

This is almost the same problem as **Problem 2** in the previous homework, however, the bounds have been changed. Fortunately that is just bounds of [0,L] to [-L,L], doubling everything. Since the inner function is made out of only cosine functions, more particuarly *even* functions, this means that the resulting funciton must be even as an even times an even is still an even. Thus, the integral must double, at least for the normal case where $m \neq n$. 

As we'll recall from **Problem 2** in the previous homework, the integral from 0 to L came out to 0, so it must be 0 on the other side as well, leaving a total integral... of zero.

Now, what of the edge cases? The m=n=0 case showcases the doubling better: the original integral was just L, but now it is 2L! This is because we are integrating over an argument of 1, which always becomes the length of the interval, which in this case is 2L. 

What of the $m=n\neq 0$ case? Well in that case we endu up with an integral of a square cosine function, which is still even, meaning we can double the integral from L/2 to get L. Or we can actually take the integral:

$$ = \left[ \frac{sin(2\frac{n\pi x}{L})}{4\frac{n\pi}{L}} + \frac x2 \right]_{-L}^L $$

Which becomes L/2 - (-L/2) = L. 

Putting all our answers together, the integral is 0 when $n \neq m$, 2L when $n=m=0$, and $L$ when $n=m\neq0$

Now we don't even have to take the integral of this one, it's an even function times an odd function, so in normal situations $n\neq m$ it goes to zero over a symmetric bound since the integrand itself is an odd function. When m=n=0 it also goes to zero since, while the cosine doesn't go to zero, the sine does, making the entire integrand zero. But, once agian, we whave to ask about $m=n\neq 0$. This time, we can't just combine the two together as the trig functions are different from one another. Using the product-to-sum formula would result in an m-n=0 situation, which we want to avoid. So, instead, let's use some actual integral solving techniques:

$$ \int_{-L}^{L} cos(n\pi x/L) sin(n\pi x/L) dx $$

One could argue that the equation is still odd and thus has to integrate to zero, but we are at an edge case and so want to be very careful. So we apply integration by parts:

$$ \int_{-L}^{L} cos(n\pi x/L) sin(n\pi x/L) dx = [\frac{L}{n\pi}sin(n\pi x/L) sin(n\pi x/L)]_{-L}^L - \int_{-L}^{L} sin(n\pi x/L) cos(n\pi x/L) dx $$

The middle term goes to zero as it's all sines of integer multiples of $\pi$, leaving us with this curious relation.

$$ \int_{-L}^{L} cos(n\pi x/L) sin(n\pi x/L) dx = - \int_{-L}^{L} sin(n\pi x/L) cos(n\pi x/L) dx $$

Since each trig function can be treated as a variable, the order of multiplication is irrelevant and we can commute. Which means that this intergral is equal to its own negative. Which is only possible if the integral is zero. 

Thus, the answer here is that the integral is 0 *with no restrictions on n and m*. (So long as they are non-negative integers).

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$



We are assuming at the start that steady-state exists, essentially that $\frac{\partial^2 u}{\partial x^2} = 0$. It's also important to recognize that $u_t=0$, even though it's implicit above, we can use it to further restrict our solutions because nothing better be changing in time. One might think that this means our ICs become our solution everywhere over all time, but that's not necessarily true, as the steady-state solution could just be a state in the distant future. So we could change at the start, so long as we eventually reach a state of no change. 

So we're going to do the perhaps dubious act of ignoring the ICs at first and just assuming it'll "flow properly." The characteristic equation for our differential equation is $r^2=0$, giving us only roots of 0, but two of them so we double up, making the solution u = A+Bx. Which, yes, makes sense, no matter what A and B are the result will be zero at the second derifative.

The first derivative conditions both require that the result be -2. Considering that $u_x = B$, that means B is -2 in the steady-state solution. Now, unfortunatley, we cannot find A...

But we can be clever about it. See, the step-by-step flow of the ICs still can't violate the conservation of heat, thus the total heat present at the start still must be present at the end. We can figure out how much there is with an integral. The first half (from 0 to 2) has no heat in it at the start, while the other half (from 2 to 4) has heat 2x-4, which integrates to $x^2-4x$, which we can evaluate to have a total "heat" of 4. This will need to be the heat at the end as well, so...

$$ 4 = \int_0^4 A-2x dx $$
$$ \Rightarrow 4 = [Ax-2x^2]_0^4 dx $$
$$ \Rightarrow 4 = 4A-32 dx $$
$$ \Rightarrow 8 = A dx $$

And so the steady state solution, when t is large, is...

$$ u(x) = 8 -2x $$

Which effectively gives us an "end of time" boundary condition for later parts of the problem. 

So let the answer to part a) be w(x). Declare the full answer to be u(x,t) = w(x) + v(x,t). 

So, what do we know about v? Well, it does NOT have to obey w's BCs, because w already solves those BCs. If the derivative is taken, w will produce -2 still, so v has to produce 0. Thus, it's BCs are zero in the x derivatives. It has to obey the ICs as, specifically through v(x,0) = u(x,0) - w(x), which is a piecewise function with 2x-8 on [0,2] and 4x-12 on [2,4]. However we do have another condition: that in the steady state, that is, as t approaches infinity, v has to approach 0 to allow w(x) to dominate. 

Now, what is v's PDE? u's is $u_t = u_{xx}$ and w's is $0 = w_{xx}$. Notably, in one of our earlier conditions we noted essentially that v = u-w. We note that if we add w to u, we do not change any of the conditions, so $v_t = v_{xx}$ is the PDE v must solve. 

We note that this is just the general heat equation with zero spatial derivatives, that is, the rod immersed in the 0 degree bath. The full solution is 2.3.30. 

$$ u(x,t) = \sum^\infty_{n=1} B_n e^{-\lambda_n \kappa t} sin (\frac{n \pi x}{L}) $$

Where $\lambda = (\frac{n \pi}{L})^2$

Of course, in our case, L is a specific value, specifically 4, and $\kappa$ is just 1. We now see why we bothered with part a), it allowed us to create v in a known solution form, that is, with the spatial derivatives at the end being zero. 

$$ v(x,t) = \sum^\infty_{n=1} B_n e^{-\lambda_n t} sin (\frac{n \pi x}{4}) $$

Now this obviously goes to zero as t tends to infinity, that exponential is going to get smaller and smaller until it poofs to nothing. However, we still need to match the initial conditions... somehow. At t=0 we need to match a piecewise function.

$$ v(x,0) = \sum^\infty_{n=1} B_n sin (\frac{n \pi x}{4}) $$

Which is to say 0 on [0,2] and 2x-4 on [2,4]. Now, the best way to do this is 2.3.35, that is, the formula to find the constants from an f(x) initial condition. With our values substitued in, of course. 

$$ B_n = \frac 12 \int^4_0 f(x) sin (\frac{n \pi x}{4}) dx  $$

Split the integral up based on the piecewise nature of the initial condition. Remember: 2x-8 on [0,2] and 4x-12 on [2,4].

$$ = \frac 12 \int^4_2 (2x-8) sin (\frac{n \pi x}{4}) dx + \frac 12 \int^4_2 (4x-12) sin (\frac{n \pi x}{4}) dx  $$

$$ = \int^4_2 x sin (\frac{n \pi x}{4}) dx - 4 \int^4_2 sin (\frac{n \pi x}{4}) dx + 2 \int^4_2 x sin (\frac{n \pi x}{4}) dx -6 \int^4_2 sin (\frac{n \pi x}{4}) dx  $$

$$ = \left[ \frac{16}{n^2\pi^2} sin (\frac{n \pi x}{4}) - x \frac{4}{n\pi} cos (\frac{n \pi x}{4}) \right]_0^2 + 4\frac{4}{n\pi} \left[ cos (\frac{n \pi x}{4}) \right]_0^2 +2\left[ \frac{16}{n^2\pi^2} sin (\frac{n \pi x}{4}) - x \frac{4}{n\pi} cos (\frac{n \pi x}{4}) \right]_2^4 + 6\frac{4}{n\pi} \left[ cos (\frac{n \pi x}{4}) \right]_2^4  $$

Now this is the part where each coefficient will be a little different, since we're evaluating from 2 to 4 we're going to end up with a lot of half-pi radian evaluations, meaning everything's going to cycle a lot with n. 

$$ = ( \frac{16}{n^2\pi^2} sin (\frac{n \pi}{2}) - 2 \frac{4}{n\pi} cos (\frac{n \pi }{2})) - 0 + \frac{16}{n\pi} \left[ cos (\frac{n \pi}{2}) - 1 \right] +2\left[ (0 - 4 \frac{4}{n\pi} cos (n \pi)) - (\frac{16}{n^2\pi^2} sin (\frac{n \pi }{2}) - 2 \frac{4}{n\pi} cos (\frac{n \pi}{2})) \right] + \frac{24}{n\pi} \left[ cos (n \pi) - cos (\frac{n \pi}{2}) \right]  $$

$$ =  \frac{16}{n^2\pi^2} sin (\frac{n \pi}{2}) - \frac{8}{n\pi} cos (\frac{n \pi }{2}) + \frac{16}{n\pi} cos (\frac{n \pi}{2}) - \frac{16}{n\pi} -  \frac{32}{n\pi} cos (n \pi) - \frac{32}{n^2\pi^2} sin (\frac{n \pi }{2}) +  \frac{16}{n\pi} cos (\frac{n \pi}{2}) + \frac{24}{n\pi} cos (n \pi) -  \frac{24}{n\pi} cos (\frac{n \pi}{2})  $$

$$ =  -\frac{16}{n^2\pi^2} sin (\frac{n \pi}{2}) - \frac{16}{n\pi} - \frac{8}{n\pi} cos (n \pi)  $$

Therefore...

$$ v(x,t) = \sum^\infty_{n=1} B_n e^{-\frac{n^2\pi^2}{16} t} sin (\frac{n \pi x}{4}) $$

$$ B_n = -\frac{16}{n^2\pi^2} sin (\frac{n \pi}{2}) - \frac{16}{n\pi} - \frac{8}{n\pi} cos (n \pi)  $$

$$ w(x) = 8 -2x $$

$$ u(x,t) = v(x,t) + w(x) $$

And all this collectively is the full solution. 

This is going to be... quite annoying to code, considering that we have an infintie series in the middle of it. Fortunately for a good plot one doesn't have to actually evaluate everything out to the infinite level, more n just means more resolution.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$



The actual separation of the variables is easy. let u = XY. Where X and Y are funcitons of one variable, x and y respectively. 

$$- \frac{\partial^2XY}{\partial x^2} = \frac{\partial^2XY}{\partial y^2}$$
$$ \rightarrow - Y \frac{\partial^2X}{\partial x^2} = X \frac{\partial^2Y}{\partial y^2}$$
$$ \rightarrow - \frac 1X \frac{\partial^2X}{\partial x^2} = \frac 1Y \frac{\partial^2Y}{\partial y^2} = -\lambda$$

Where $\lambda$ is some constant. Note that we are NOT completely symmetric and x and y. The "official" separation in this case would be:

$$  - \frac 1X \frac{\partial^2X}{\partial x^2} - \lambda = 0$$
$$  \frac 1Y \frac{\partial^2Y}{\partial y^2} - \lambda = 0$$ 

It will be easier to solve them in an alternate form.

$$  - \frac{\partial^2X}{\partial x^2} - \lambda X = 0$$
$$  \frac{\partial^2Y}{\partial y^2} - \lambda Y = 0$$ 

Both of these are simple second-order differential equations. We can solve with the auxilliary polyn9omials. For x this is $r^2-\lambda=0$ and for y this is $-r^2-\lambda=0$. These can be split up into $(r-\sqrt\lambda)(r+\sqrt\lambda)=0$ and $-(r+i\sqrt\lambda)(r-i\sqrt\lambda)=0$ respectively. This makes our solutions...

$$ X = Ae^{-i\sqrt\lambda x} + Be^{i\sqrt\lambda x} = Acos(\sqrt\lambda x) + Bsin(\sqrt\lambda x) $$
$$ Y = Ce^{-\sqrt\lambda y} + De^{\sqrt\lambda y} $$

Note that the A and B constants do not strictly go across the equality, but they are currently completely arbitrary so there's no need to define extra constants to differentiate them. 

Now, what do the boundary conditions tell us? 

u(0,y)=0 implies X(0)Y(y) = 0. Since we are looking for nontrivial soluitons, Y(y) can't equal zero, so X(0)=0 must be true. 

u(L,y)=0 implies X(L)Y(y)=0. Similarly to the above, this means X(L)=0. 

u(x,0)=0 implies X(x)Y(0)=0. Again, this means Y(0)=0, as X(x) has to be nontrivial. 

u(x,H)=f(x) implies X(x)Y(H) = f(x). This is not as helpful as the previous results, since we don't know what f(x) is. 


Now we can see why we set the minus sign on the X separation of variables: we needed that boundary condition to be periodic, that is, to be 0 at two different points. We do not do that with Y because Y is not a homogenous boundary value problem, so our equations for it might not even be valid, it remains to be seen if it will be useful. 

Since x=0 X=0, we have to have a sine function and not a cosine, thus A=0. At this point it should be pretty clear that $\lambda = (\frac{n\pi}{L})^2$ in order to match these boundaries, where n is a positive integer of somekind. Of course, we still need to justify that there are no zero or negative solutions first... but wait! If $\lambda$ were negative, the square root would prodice imaginary numbers. If that occurs, the solution for X becoems composed of ordinary exponentials, which cannot be solved to be zero in two different places, so there are no negative solutions. As for zero, if $\lambda=0$, X becomes a constant. A constant only solves our equation if it is zero, which is the trivial solution, which we do not care about at this juncture.

So we have our solution for X:

$$ X(x) = B sin(\frac{n\pi x}{L}) $$

B is currently an arbitrary constant that we can just absorb into Y, since u = YX, and removing the constant does not destroy our definition of X. 

Speaking of Y, we now know $\lambda$.

$$  \frac{\partial^2Y}{\partial y^2} - (\frac{n\pi}{L})^2 Y = 0$$ 

Which is a known form with a solution 2.5.25

$$ Y = Ccosh(\frac{n\pi}{L}y) + Dsinh(\frac{n\pi}{L}y) $$

Though rather obviously the cosh term vanishes since it can't satisfy Y(0)=0. So we end up with...

$$ Y(y) = Dsinh(\frac{n\pi}{L}y) $$

Where D is still our arbitrary constant. Since u=YX we declare that D also contains B for simplicity's sake. Regardless, this gives us the total solution:

$$ u(x,y) = D sin(\frac{n\pi x}{L}) sinh(\frac{n\pi}{L}y) $$

Of course we ahve only solved three of the conditions, the 0 conditions. We have a fourth condition based on f(x). Fortunately we can use superposition to add all possible combinations together and solve with that. Replace D with $A_n$ for consistency of notation. 

$$ u(x,y) = \sum_{n=1}^\infty A_n sin(\frac{n\pi x}{L}) sinh(\frac{n\pi}{L}y) $$

Now, let u=f(x) at y=H. 

$$ f(x) = \sum_{n=1}^\infty A_n sin(\frac{n\pi x}{L}) sinh(\frac{n\pi H}{L}) $$

Note that the sinh function is no longer a function, but a constant, and thus can be considered part of $A_n$ in the full expansion. So the question becomes how do we determine these coeffeiceints? The properties of orthogonality. Luckily this has already been done once in 2.3.35 for anything of the basic sine form. 

$$ A_n sinh(\frac{n\pi H}{L}) = \frac{2}{L} \int^L_0 f(x) sin\frac{n\pi x}{L} dx $$
$$ \Rightarrow \boxed{A_n = \frac{2}{Lsinh(\frac{n\pi H}{L})} \int^L_0 f(x) sin\frac{n\pi x}{L} dx} $$

Naturally to go any further would require knowing what f(x) actually is. (note that the bound is L rather than H. This is becuase f(x) is the boundary that takes place AT y=H, leaving it to vary from 0 to L).

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$



Visually, this is a "pizza slice" with three bounds--the two flat edges have a heat level 0, and the curved end follows some funciton $f(\theta)$. This also means it's zero at the origin, as both of the 0 boundary conditions converge there.  Also since radians are genreally better to work with, we note that 60 degrees is $\frac \pi3$ radians. 

Laplace's equation in polar coordinates is:

$$ 0 = \frac 1r \frac{\partial}{\partial r} (r \frac{\partial u}{\partial r}) + \frac{1}{r^2} \frac{\partial^2 u}{\partial \theta^2} $$

Which rather quickly offers a separation of variables situation. Let $u(r,\theta) = R(r)\Theta(\theta)$. 

$$ \Rightarrow - \frac 1r \frac{\partial}{\partial r} (r \frac{\partial R\Theta}{\partial r}) = \frac{1}{r^2} \frac{\partial^2 R\Theta}{\partial \theta^2} $$

$$ \Rightarrow - \frac{1}{rR} \frac{\partial}{\partial r} (r \frac{\partial R}{\partial r}) = \frac{1}{r^2}\frac{1}{\Theta} \frac{\partial^2 R\Theta}{\partial \theta^2} $$
$$ \Rightarrow - \frac{r}{R} \frac{\partial}{\partial r} (r \frac{\partial R}{\partial r}) = \frac{1}{\Theta} \frac{\partial^2 R\Theta}{\partial \theta^2} $$

Which is now officially separated. We can state the actually separated equaitons:

$$ - r \frac{\partial}{\partial r} (r \frac{\partial R}{\partial r}) + \lambda R= 0$$
$$  \frac{\partial^2 R\Theta}{\partial \theta^2} + \lambda \Theta = 0$$

Now $\theta$ is the variable with the neater boundary conditions, both in the fact that they're zero and the fact that the separated equation is a lot less of a mess. 

Now there's some thing we know immediately from previous work. First of all, the angular dependence is periodic in exactly the same way as the Laplace equation was in **Problem 4** for X. We already know this oslution. 

$$ X(x) = B sin(\frac{n\pi x}{L}) $$

In our variable...

$$ \Theta(\theta) = B sin(3n \theta) $$

Since we know what L is. This is, naturally, for positive $\lambda$. Negative solutions cannot satisfy our periodicity, and zero solutions are trivial. (NOTE: This is only the case because our boundary conditions have to go to zero! If we were on a full disc, not only would we need to consider zero, we would also need to consider the cosine solutions as well). Here, $\lambda$ specifically equals $9n^2$. It's nice that our "length" L has a term of $\pi$ in it. 

Anyway, now let's examine the radial equation. We can simplify it somewhat with derivative rules.

$$ - r^2 \frac{\partial^2 R}{\partial r^2} - r \frac{\partial R}{\partial r} + \lambda R= 0$$

Now we pull out a fancy trick--assume $R = r^p$, where p is some constant. If we do this, we can force out a characteristic equation. (In cartesian we usually can't do this with variables outside the derivatives, but here it just so happens to work.) If we work out the derivatives, we get

$$ 0 = -r^2 p (p-1) r^{p-2} - r p r^{p-1} + \lambda r^p $$

Which we can combine like terms to get the characteristic equation.

$$ 0 = r^p (-p(p-1) - p + \lambda) $$

Now since we assume the solution isn't trivial, this means $ -p(p-1) - p + \lambda = 0$. This simplifies rather quickly to $p^2 = \lambda$, which means $p = \pm 3n$

Which means we can actuall write down the general equation for R.

$$ R = Cr^{3n} + Dr^{-3n} $$

Note that the zero $\lambda$ solution produces an R=E result, where E is some constant. This, while simple, is not trivial so it qualifies as one of our solutions. Which, yes, obviously a constant satisfies the laplacian, but it does NOT satisfy our boundary conditions (unless that constant is zero). Now, we need to avoid exploisons at the origin, which means D=0 and thus...

$$ R(r) = Cr^{3n} $$

Which means we have a solution for our total...

$$ u = R\Theta = C r^{3n} sin(3n\theta) $$. 

Where C is the consolidation of all the leading constants. Naturally, to get the full solution we apply superposition. 

$$ u = \sum_{n=1}^\infty A_n r^{3n} sin(3n\theta) $$

Now, we still need to find $A_n$, and that's what our last boundary condition is for, when r=a, u=f(x). 

$$ f(x) = \sum_{n=1}^\infty A_n a^{3n} sin(3n\theta) $$

a is now just a constant, so it can be solved *with* A_n, just like we did in **Problem 4**. 

$$ A_n a^n = \frac{6}{\pi} \int^{\pi/3}_0 f(\theta) sin(3n\theta)d\theta $$
$$ \Rightarrow \boxed{A_n = \frac{6}{\pi a^n} \int^{\pi/3}_0 f(\theta) sin(3n\theta)d\theta } $$
