# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$



Induction means "prove it for the initial case and then prove that if it is the case for a previous iteration, it must be the case for the latter iterations."

The first iteration, in this case, is M=1. Here, there is no sum at all, and thus the relation we try to prove becomes:

$$ \mathcal{L} (c_1u_1) = c_1 \mathcal{L} u_1 $$

Which is trivially true since constants can move in and out of operators at will. Thus, it works for M=1.

Now, let us examine the case for M. 

$$ \mathcal{L} (\sum^M_{n=1} c_nu_n) = \sum^M_{n=1} c_n \mathcal{L} u_n $$

We split off the Mth term from both sides.

$$ \Rightarrow \mathcal{L} (\sum^{M-1}_{n=1} (c_nu_n) + c_Mu_M)  = \sum^{M-1}_{n=1} (c_n \mathcal{L} u_n) + c_M \mathcal L u_M $$

Linear operators obey their version of the distributive property, so long as we keep it "left-justified."

$$ \Rightarrow \mathcal{L} \sum^{M-1}_{n=1} (c_nu_n) + \mathcal{L} c_Mu_M  = \sum^{M-1}_{n=1} (c_n \mathcal{L} u_n) + c_M \mathcal L u_M $$

As we did for the M=1 case, move the constant out. 

$$ \Rightarrow \mathcal{L} \sum^{M-1}_{n=1} (c_nu_n) + c_M \mathcal{L} u_M  = \sum^{M-1}_{n=1} (c_n \mathcal{L} u_n) + c_M \mathcal L u_M $$

Now both sides have the same term so cancel. 

$$ \Rightarrow \mathcal{L} \sum^{M-1}_{n=1} (c_nu_n)  = \sum^{M-1}_{n=1} (c_n \mathcal{L} u_n)$$

What we have done here is extract the M-1 case from the M case, which means we can do that for every possible value of M. From M-1 we can extract M-2, and so on and so forth until we arrive at M=1. We know M=1 makes the relation true. Therefore, we can use the M=1 case to construct every other case, no matter what M is.

This is the principle of mathematical induction. If M=1 is true AND if M-1 implies M (or M implies M+1) then M is true. We have both criteria met, so the given relation is true.

$$ \int_0^L cos\frac{n\pi x}{L}cos\frac{m\pi x}{L} dx $$

$$ = \frac 12 \frac{L}{(n+m)\pi} sin((n+m)\pi) + \frac 12 \frac{L}{(n-m)\pi} sin((n-m)\pi) $$

As previously mentioned any integer multiple of $\pi$ in sine goes to zero, leaving only the fractional term at the end.

Thus the integral for $m = n \neq 0$ is **L/2**

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$



This is clearly a heat equation, where u is the funciton of heat on a rod of some kind, with a given starting heat distribution of f(x) and boundary conditions that demand the heat at locations x=0 and x=L are "flat" in space as their spatial derivatives are zero. 

It is possible that this doesn't actually represent heat, as the heat equation has many uses, but heat is a valid physical interpretation. 

How about first we actually separate the variables? We can write our differential equation as $u_t = \kappa u_{xx}$ which, if we take u = X(x)T(t) = XT, we can split up...

$$ (XT)_t = \kappa (XT)_{xx} $$
$$ \Rightarrow XT_t = \kappa TX_{xx} $$
$$ \Rightarrow \frac 1T T_t = \kappa \frac 1X X_{xx} $$

Which gives us the two equations

$$ \Rightarrow \frac{1}{T\kappa} T_t +\lambda = 0 $$
$$ \frac 1X X_{xx} + \lambda = 0 $$

Notice that we moved the $\kappa.$ This is to save us a headache later. 

Now that everything's actually separated, we can actually take a step back to see where we go from here. It will be helpful to explicitly state all the boundary conditions for both the X and T sections as they follow from the above. First of all, u(x,0) = f(x) tells us that X(x)T(0) = f(x). Which admittedly tells us nothing aside from the fact that T(0) is a constant, which it would have to be anyway. If we were given f(x) that would certainly narrow down the options for X, as X(x) = f(x)/T(0), but we are not. (That said, the fact that u(x,0) = f(x) implies heavily that we are justified in saying u = XT)

The boundary conditions in this problem, we note, are difrential ones, NOT direct relations (which is why we didn't describe the problem as a rod with ends of fixed temperature. The temperature sure can change, it's just that the SPATIAL derivative has to be zero there). It still tells us some information:

$$T(t)\frac{\partial}{\partial x}X(0) = 0$$
$$T(t)\frac{\partial}{\partial x}X(L) = 0$$

The obvious solution to these equation, T(t)=0, is considered a "trivial" one as that sets everything to zero, which is effectively useless. So T(t) has to be some nonzero constant, meaning that the equations give an even greater restriction:

$$\frac{\partial}{\partial x}X(0) = 0$$
$$\frac{\partial}{\partial x}X(L) = 0$$

Which makes sense with our original understanding: the temperature can change on the ends through time, yes, but not through space. There are a few ways X could satisfy these equations though: X itself could be a constant, or X could be one of those functions with extrema at 0 and L -- for instance, a cosine derivative function. 

Since the T portion of the problem is closer to the standard, let's tackle it first. 

$$ \frac{1}{T\kappa} T_t +\lambda = 0 $$

This is actually a known form, a simple exponential.

$$ T(t) = Ce^{-\lambda \kappa t} $$

where notably $T_t / T$ does in fact become $-\lambda$ and so everything is satisfied. The question now is what C is. Now, since we do not expect heat to grow exponentially in time, we suspect $\lambda \geq 0$, but we cannot prove this yet like the problem wants. But we do have the condition, which is good, find $\lambda$, determine the exponentiality of the equation. As for C, it's just a leading constant, it can be absorbed into other things later when we apply boundary conditions. After all, one may note that we don't really have boundary conditions on T... The initial condition simply says that at t=0, T is a constant. Which it is, and that constant is C. Exactly what constant that is depends somewhat on f(x) and X. 

Now we return to the X equations.

$$ \frac 1X X_{xx} + \lambda = 0 $$
$$\frac{\partial}{\partial x}X(0) = 0$$
$$\frac{\partial}{\partial x}X(L) = 0$$

Currently, $\lambda$ is arbitrary. We will make the assumption that it is not complex, but right now it could be positive, negative, or zero. (Notably we expect to remove the negative values as that would cause exponential exploding, but we want to *show* that.)

Let's try to remove the negative values first. Notably the form of our actual differential equation is identical to that of the formulation on page 41 of the textbook, so we know what the general answer is already. Notably, the boundary conditions are NOT the same. 2.3.25 gives us the general solution.

$$ X = C_5 cosh (\sqrt{s}x) + C_6 sinh (\sqrt{s}x) $$

Where s is $-\lambda$ for convenience and not looking like we're taking the square root of a negative. 

We started out by arguing that sinh and cosh had zero and one zero-derivative points and had no way to combine to make there be *two*, which would be required for our boundary conditions. But there's an even simpler proof than that. sinh and cosh are each other's derivatives. We already know that cosh and sinh functions cannot solve the same boundary conditions *without* the derivatives in them. Since adding the derivatives doesn't change the constituent functions, it *still* can't work! (Page 42 is where this is shown).

So there are no negative $\lambda$ solutions. This also does what the problem suggested we do: show that there are no exponentially growing solutions. 



Now what if $\lambda$ was zero? Again, the solution to X is known, given on page 41 as an unlabaled equation (possibly because it's so well known.)

$$ X = C_1 + C_2x $$

We need the derivative of this to apply the boundary conditions. X' = C_2. Which is only true if C_2 is 0. However, this does leave open a possibility that C_1 actually equals something. 

$$ X = C_1 $$

This seems like it will correspond with the $A_0$ in the given "goal" in the problem statement, but it could be combined with another constant. The specific value of this constant would be related to f(x). 

And now we ask the most interesting quesiton, what if $\lambda$ were positive? The general solution to the differential equation is 2.3.18

$$ X = C_3 cos(\sqrt \lambda x) + C_4 sin(\sqrt \lambda x)$$

Now this one easily fits our differential boundary conditions, so long as we can get the trig funcitons to line up. Since we are using the derivatives to measure the boundary, that means $C_4$ has to be zero as sin's derivative is cos, and that's can't be zero at zero. However, C_3 can exist, as the derivative of cos most certainly *can* be zero at both zero and L. 

$$ X = C_3 cos(\sqrt \lambda x)$$

Now the derivative of this being 0 at 0 is true always, but being 0 at L requires us to define $\lambda$ in such a way that the zero of the trig function lands on L. This actually isn't hard.

$$ \lambda = \left(\frac{n\pi}{L}\right)^2 $$

No matter what, when x is set to L, this will result in an integer multiple of $\pi$ being inserted into the cosine function. Which will make it vanish, exactly what we want. 

$$ X = C_1 $$
$$ X = C_3 cos(\sqrt \lambda x)$$
$$ \lambda = \left(\frac{n\pi}{L}\right)^2 $$
$$ T = C_0e^{-\lambda \kappa t} $$

Note that we called the T-constant C C_0 here to differentiate it from the others. 

Techncially, these equations can be combined to create a solution, one for each value of n, treating the $C_1$ case as n=0. However, given that the answer is put in the form of a sum, that likely isn't going to cut it and we'll need to rely on superposition, that is, we're going to write it all down at once to represent the linear combination of every solution. 



When we do this, we combine all the leading constants together into one big A term. The initial result is:

Note that $A_0$ is literally just $C_1C_0$. 

Now, if we expand one of the $\lambda$ terms, we get the answer provided by the problem statement.

Notably there's a k instead of a $\kappa$ but we think that's just a typo in the problem statement. The subscript n on $\lambda$ indicates that it changes with n. 

However, we are not quite done. We found $\lambda$, but our constants $A_n$ are still unknown. How are we going to find it? Its constituent constants are $C_0C_3$, but those were arbitrary constants too! This is where we turn to the mysterious and oft-alluded to f(x).

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$



For the first one, note that r = $\sqrt{x^2+y^2}$. The square root does not actually provide any ambiguity, since r itself has to be positive, being a radial measurement. We can take the derivative directly for both x and y, producing 

$$ \frac{\partial r}{\partial x} = \frac{x}{\sqrt{x^2+y^2}} = \frac xr = cos\theta $$
$$ \frac{\partial r}{\partial y} = \frac{y}{\sqrt{x^2+y^2}} = \frac yr = sin\theta $$

Now we know the arc trig functions have limited domains, so we simply cannot use them here. So how are we going to take a derivative of $\theta$? Well, what if we didn't do it directly. Say we take the derivative of $sin\theta$ instead. 

$$\frac{\partial cos\theta}{\partial y} = -sin\theta \frac{\partial \theta}{\partial y}$$

Now we know what $cos\theta$ is, so we can make that substitution. 

$$\Rightarrow \frac{\partial}{\partial y} \frac{x}{\sqrt{x^2+y^2}}= -sin\theta \frac{\partial \theta}{\partial y}$$
$$\Rightarrow \frac{-xy}{(x^2+y^2)^{3/2}}= -\frac yr \frac{\partial \theta}{\partial y}$$
$$\Rightarrow \frac{x}{r^2}= \frac{\partial \theta}{\partial y}$$
$$\Rightarrow \frac{cos\theta}{r}= \frac{\partial \theta}{\partial y}$$

And the other way we get

$$\frac{\partial sin\theta}{\partial x} = cos\theta \frac{\partial \theta}{\partial x}$$
$$\Rightarrow \frac{\partial}{\partial x} \frac{y}{\sqrt{x^2+y^2}}= cos\theta \frac{\partial \theta}{\partial x}$$
$$\Rightarrow \frac{-xy}{(x^2+y^2)^{3/2}}= \frac xr \frac{\partial \theta}{\partial x}$$
$$\Rightarrow \frac{-y}{r^2}= \frac{\partial \theta}{\partial x}$$
$$\Rightarrow \frac{-sin\theta}{r}= \frac{\partial \theta}{\partial x}$$

And that proves all four of the relations. 

The first part is a basic property of vector algebra. The projection of any vector onto another vector is the angle between those two vectors times the vector being projected onto; the projection of $\hat r$ onto $\hat i$ is thus $cos\theta \hat i$. The projection onto $\hat j$ would involve the inverse angle... or just the sine, $sin\theta \hat j$. Since $\hat j$ and $\hat i$ are orthogonal, projections onto both of them constitute a complete decomposition of the vector.

Now, what of the radial component? The radial unit vector is always perpendicular to the radius r, so a simple transformation from $\hat r$ to $\hat \theta$ will do it, specifically a rotation of 90 degrees. Changing sines to cosines and cosines to sines accomplish this. However, we need to concern ourselves with direction--the direction of $\hat \theta$ is counterclockwise, so when the angle is zero it needs to point straight up and then 90 degrees later point directly to the negative x direction. All these criteria are met by $-sin\theta \hat i + cos\theta \hat j$ and only that. (Any constants added would make it not a unit vector.) 

Let us start in cartesian.

$$ \nabla = \hat i \frac{\partial}{\partial x} + \hat j \frac{\partial}{\partial y} $$

We need to convert this into polar. 

From Marsden and Tromba's *Vector Calculus* Ex 2.5.4 we see that we don't need to convert this directly and can, as the problem suggested, just use the chain rule. However, this is the more general form of the chain rule used on differentials that we regularly forget exists. 

$$ \frac{\partial}{\partial x} = \frac{\partial r}{\partial x} \frac{\partial}{\partial r} + \frac{\partial \theta}{\partial x}\frac{\partial}{\partial \theta} = cos\theta  \frac{\partial}{\partial r} + \frac{-sin\theta}{r} \frac{\partial}{\partial \theta} $$

And similarly for the y corodinate. 

$$ \frac{\partial}{\partial y} = \frac{\partial r}{\partial y} \frac{\partial}{\partial r} + \frac{\partial \theta}{\partial y}\frac{\partial}{\partial \theta} = sin\theta  \frac{\partial}{\partial r} + \frac{cos\theta}{r} \frac{\partial}{\partial \theta} $$

These can be substituted into the gradient operator. 

$$ \nabla = \hat i (cos\theta  \frac{\partial}{\partial r} + \frac{-sin\theta}{r} \frac{\partial}{\partial \theta}) + \hat j (sin\theta  \frac{\partial}{\partial r} + \frac{cos\theta}{r} \frac{\partial}{\partial \theta}) $$

The relations in part b) can more or less be applied directly at this point, note that the first terms in the parentheses are part of $\hat r$ and the second terms $\hat \theta$

$$ \Rightarrow \nabla = \hat r \frac{\partial}{\partial r} + \hat \theta \frac{1}{r} \frac{\partial}{\partial \theta} $$

Which is exactly what we sought. 

Rather than taking any chances we're going to assume those derivatives aren't proven yet and that we have to prove them. However, it is rather obvious. As the unit vectors $\hat i$ and $\hat j$ are never changing, they are unaffected by any kind of derivative, so it's only the angles that are altered. If one examines b), one can clearly see that the derivative of $\hat r$'s components is $\hat \theta$, and that taking the derivative again on trig functions just adds a minus sign. So the derivatives do follow rather quickly. 

So now, equipped with our new gradient operator, we introduce its divergence by simply applying it and seeing what happens. Note that we have to be careful, the dot product is not defined as simply as it is in cartesian coordinates. Now normally to resolve this we would just apply the metric (diagonal 1,$r^2$) and evaluate the dot product that way, but the gradient operator is insidiously not actually a vector but a one-form, so we need to be *very clear* in what exactly this dot product means. 

Unfortunately all we could find in the differential equations book was the definition for cartesian, which is decidedly unhelpful becuase cartesian does not have the nuances of other coordinate systems. 

$$ \nabla \cdot \vec A = \frac{\partial}{\partial x} A_x + \frac{\partial}{\partial y}A_y $$

So let's try it completley generally, which is to say, treat this as an outer product where every term gets multiplied, like two binomials.

$$ \left( \hat r \frac{\partial}{\partial r} + \hat \theta \frac{1}{r} \frac{\partial}{\partial \theta} \right) \cdot \left( A_1 \hat r + A_2 \hat \theta \right) $$
$$ = \left( \hat r \frac{\partial}{\partial r} \cdot A_1 \hat r + \hat r \frac{\partial}{\partial r} \cdot A_2 \hat \theta + \hat \theta \frac{1}{r} \frac{\partial}{\partial \theta} \cdot A_1 \hat r + \hat \theta \frac{1}{r} \frac{\partial}{\partial \theta} \cdot A_2 \hat \theta  \right) $$

One may ask why the dot products are still there. It is because we are being extremely explicit, the unit vectors *need* to still dot with each other so they can cancel or do other things. However, these new dot products only care about the unit vectors, as they are the only vectors left, even the differentials can be treated as constants. Every differential, though, is acting on two things. The results will arise from the product rule. Derivatives with respect to r are just zero, since none of the components depend on r. We admit we thought we were working with spherical symmetry and not circular for quite a while, which would definitely have a different answer ($4 \pi$), but the principles are the same. 

Note that the integration factor for polar coordiantes is $r drd\theta$. This can be derived in a few ways, but our presonal favorite is taking the square root of the trace of the metric, since that can be done in all coordinate systems and doesn't bog the brain down. (The polar metric has a diagonal of 1, $r^2$, so a trace of $r^2$, and hey look there's the r). 

Anyway, for any circular situation we have an area integral. For all thermal energy, it is set up as:

$$ \int\int_R c \rho u dA $$

Which, in our case, is

$$ \int_a^b \int_0^{2\pi} c \rho u r d\theta dr $$ 

Nothing depends on the angle so we integrate over the angle like a constant, getting

$$ 2\pi \int_a^b c \rho u r dr $$ 

Which is exactly what we were looking for. 

Essentially we need to derive a new form of Fourier's Law of HEat Conduction. $K_0$ is the thermal conductivity. The 1D version of this equation is 1.2.8 $\phi = -K_0 \frac{\partial u}{\partial x}$. Now logically the result has to be the one above, because the heat flow is radially symmetric and thus is going to be spread evenly across a circumference of $2\pi r$ where r=b at our specific part. 

For higher dimensions, the law of heat confduction is 1.5.6, $\phi = -K_0 \nabla u$. From **Problem 5** we have

$$ \Rightarrow \nabla = \hat r \frac{\partial}{\partial r} + \hat \theta \frac{1}{r} \frac{\partial}{\partial \theta} $$

Now, we know u does not depend on the angle, so that term vanishes, meaning $\nabla u = \hat r \frac{\partial}{\partial r} u$. This, however, is just the heat conduction at any given point. We have to integrate over the entire boundary in all r directions to add it all up, and that boundary is $2\pi r$, giving us:

$$ \phi = -2\pi r K_0 \frac{\partial u}{\partial r} $$

This formula is for any radius. If we want it at a specific one, say b, we evaluate at b. 

$$ \phi = \left. -2\pi b K_0 \frac{\partial u}{\partial r}\right]_{r=b} $$

Let's apply some logic here. The flux is the rate of change of the total heat. So we just... take the time derivative of the integral and set it equal to the heat lost! Also, c and $\rho$ are constants in the new approximation, and we're solving for the general case so a=0 and b=r. 

$$ \frac{\partial}{\partial t} 2\pi c \rho \int^r_0 ur' dr' = - 2\pi r K_0 \frac{\partial u}{\partial r} $$

The trick is to take the r derivative of both sides to get rid of that pesky integral.

$$ \Rightarrow \frac{\partial}{\partial t} 2\pi c \rho u r  = -\frac{\partial}{\partial r}(2\pi r K_0 \frac{\partial u}{\partial r}) $$
$$ \Rightarrow 2\pi c \rho r \frac{\partial u}{\partial t}  =-2 \pi K_0 \frac{\partial}{\partial r} (r \frac{\partial u}{\partial r}) $$
$$ \Rightarrow \frac{\partial u}{\partial t}  = - \frac{K_0}{c \rho} \frac{1}{r} \frac{\partial}{\partial r} (r \frac{\partial u}{\partial r}) $$
$$ \Rightarrow \frac{\partial u}{\partial t}  = \frac{\kappa}{r} \frac{\partial}{\partial r} (r \frac{\partial u}{\partial r}) $$

And we're done! $\kappa$ is an arbitrary constant anyway, so it could easily absorb any constants and signs. 

The equilibrium temperature distribution occurs when the time derivative equals zero. This lets us adapt the result from part c)

$$ 0 = \frac{\kappa}{r} \frac{\partial}{\partial r} (r \frac{\partial u}{\partial r}) $$

So now let's split this up and actually solve the differential equation.

$$ \Rightarrow 0 = \frac{\kappa}{r} (r u_r)_r $$
$$ \Rightarrow 0 = \frac{\kappa}{r} (r u_{rr} + u_r) $$
$$ \Rightarrow 0 = u_{rr} + \frac 1r u_r $$

Simplifying...

$$ \Rightarrow 0 = ru_{rr} + u_r $$

First of all, we can easily reduce the order here and solve for that.

$$ \Rightarrow 0 = rw_{r} + w $$

This is a first order differential equation, so we can solve it with the first order formula. 

$$ y = \frac{\int \mu(t) g(t) dt + C}{\mu(t)} $$

where

$$ \mu(t) = e^{\int p(t) dt}$$

Now in our case it is homogeneous so g(t)=0. p(t) actually takes 1/r, and all the t variables are r variables. 

This makes $\mu(r) = e^ln|r| = |r|$ of course, r is always positive, so this is just $\mu(r)=r$

The full relation is then $$ w = \frac{C}{r} $$

Which, if we integrate, gives us...

$$ u = \frac Cr + D $$

Which is sensible, a 1/r solution and a constant solution. 

Now we get to apply our boundary conditions!

The boundary at b is insulated, so u cannot be losing any heat there, meaning that $u_r = 0$ at b. a is being held at a fixed temperature, so it also has $u_r = 0$ but it ALSO has u(a) = const. 

Note that $u_r = C/r$. There is no way for this to be 0 at both a and b unless C=0. Which thus means that u *has* to be constant everywhere. Which means we get the rather boring result of u=D, which. Which means that in equilibrium, if the temperature of a is held constant at T, the temperature of the entire region will eventually reach T. u=T. Everywhere. 

At first this seems much more like a source at a providing heat--however, all we are told is that, once we are in equilibrium, a is at T. Presumably in the past it was much hotter and then the heat distributed out! The outer edge is just insulated, meaning heat can't pass through, so there is literally no reason the heat will not evenly distribute itself across the area.
