# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

There are probably several ways to do this, but the one we think is simplest is the Taylor Expansion. This is a very well known series for $e^x$

$$ e^x = \sum_{n=0}^\infty \frac{x^n}{n!} $$

Each individual term of this series is eitehr an even or an odd function, corresponding to the even-ness or the odd-ness of the value of n. All the odd n values have odd terms, and the even n values have even terms. (This includes the n=0 case, for f(x)=1 is still an even function, albeit an unusual one). 

An even funciton plus an even function is an even function. So all the even functions added together are still even.

An odd function plus an odd function is an odd function. So all the odd functions added togetehr are still odd.

Thus...

$$ e^x = \sum_{n=0,2,4...}^\infty \frac{x^n}{n!} + \sum_{n=1,3,5...}^\infty \frac{x^n}{n!}  $$

Is clearly a construction made from an even and an odd funciton. 

Now if we wanted to apply this knowledge to an arbitrary function f(x), well we can still make it work in virtually any case via the Fourier transform. The Fourier sine series is odd, the Fourier cosine series is even, and thus every function where the Fourier series functions can be made up of an even and an odd function.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

The even part of f(x) is $f_e(x) = \frac12 [f(x) + f(-x)]$

The extensions are a little different. They take a function, f(x), and take the portion of it from 0 to L (let L approach infinity in this case), and then proceed to completely ignore the left-hand side opting to forcibly reflect the function on the 0 to L section to the -L to 0 section. We can see this in play on pg 101 for the odd extension. 

This means we are essentially defining the extension as a piecewise function. In the case of even, it is f(x) when x>0 and f(-x) when x<0--or we could alternatively write it as f|-x|. The book itself writes out the even version as

$$ f(x) \text{ when } x>0 $$
$$ f(-x) \text{ when } x<0 $$

Note that both these relations do include both f(x) and f(-x) in them, however the very important difference is that the extension is entirely ignoring the negative half of the domain, opting to just overwrite it. Also note the lack of the $\frac12$ scaling factor on the extension. 

The odd part of f(x) is $f_o(x) = \frac12 [f(x) - f(-x)]$

The extension is much the same as the above one, except the relation is -f(-x) or

$$ f(x) \text{ when } x>0 $$
$$ -f(-x) \text{ when } x<0 $$

The comparisons between the even extension and part are the same as the odd extension and part. 

First of all, we note the rather drastic split down the middle. The extensions are going to ignore the $x^2$ term entirely, while the "parts" are going to take both into account. Yet, in the end, we predict that both will be accurate in the positive domain.

To properly add things, we have to find what they actually are.

The even part for the positive domain is $ \frac12[x+x^2]$

The even part for the negative domain is $\frac12[x^2-x]$

The odd part for the positive domain is $\frac12[x-x^2]$

The odd part for the negative domeain is $\frac12 [x^2+x]$

Naturally, when we add the even and odd parts back together, we end up with x in the positive domain and $x^2$ in the negative domain. Which is to say, **we recover the original function**. As we should have. This is not going to happen with the extensions. 

It's a little odd that we're being asked to do this graphically when everything is decidedly piecewise, but we'll generate an image anyway. 



Even is purple, odd is orange, though that should be graphically obvious. When we add them, we get:



Which is the original function. 

Let's look at our extensions:

Even:

$$ x \text{ when } x>0 $$
$$ -x \text{ when } x<0 $$

Odd:

$$ x \text{ when } x>0 $$
$$ x \text{ when } x<0 $$

(the odd version could just be stated as x, which is sensible, as x itself is an odd function.)

Now, when we add thsese, the result becomes:

$$ 2x \text{ when } x>0 $$
$$ 0 \text{ when } x<0 $$

The "extended" parts completley cancel, as expected. Now, we also have a difference in the positive domain: it's 2x, not x. That's because the extensions were not trying to "split up" anything, they were trying to match the function exactly. Thus, when they were added, they doubled it. This is why the even and odd parts have the factor of $\frac12$ in front of them.

Visually, this is



Orange is odd, purple is even, note how they perfectly overlap on the positive domain. 



Added together, the negative domain entirely vanishes, while the positive domain is twice as large as the original function. Now if we added the 1/2 scaling factor we would still have an accurate reading of the positive domain, which is essentially how Fourier Series work.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

Technically speaking, (1) probably shouldn't have an equals sign, but rather a ~ since there is no guarnatee that the Fourier Series converges to the function. However, $e^x$ is a continuous function so there's no reason to think it won't so we shall continue to assume this is correct.

The genreal form of the cosine series is equation 3.3.18

$$ f(x) \text{~} \sum_{n=0}^\infty A_n cos\frac{n\pi x}{L}$$

Which, if we assume that the Series converges, does in fact reduce to:

$$ e^x = A_0 + \sum_{n=1}^\infty A_n cos\frac{n\pi x}{L}$$

Since when n=0 the cosine turns to 1. 

Now, though, we ask if we can differentiate. Differentiating series term-by-term can cause issues--even if the function itself converges, its derivative might not. We note immediately that the first and second derivative gain coefficient-n related terms that explode as n gets larger, which is a problem. However, it is not necessarily a deal-breaker.



We can clearly see $e^x$ and $e^{-x}$ meet at a sharp point which makes the derivative have a jump discontinuity. Which means we can't just take the series derivative. However, this is still fine for the cosine series. The only requirement we have for doing term-by-term differentiatino is that f'(x) is piecewise smooth. And the derivative, while it has a jump disconitnuity, is. So we CAN, in fact, take the first derivative. 

But what about the second? Well, the original step here notes that the first derivative is the sine series of $e^x$. This is true, as the derivative of $e^{-x} = -e^{-x}$, so the negative domain flips perfectly when the derivative is taken. However, we want to take this derivative AGAIN. However, the sine series has another criteria: f(0)=0 and f(L)=0. Which, as we can tell from the above, is obviously not true no matter what, derivative or no. So the third step cannot be taken and we cannot get a second derivative, and thus cannot make the relation the problem said we could 

However, we still need to use this technique to find the actual values of the constants, just proving that "here's where it's all wrong" is not sufficient. Fortunately, equation 3.4.13 tells us something about what we can do. Even if we can't differentiate it term-by-term, it IS possible to find a relation for the derivative of the sine series.

$$ f'(x) \text{~} \frac1L [f(L) - f(0)] + \sum_{n=1}^\infty \left[ \frac{n \pi}{L}B_n  + \frac2L ((-1)^n f(L)-f(0))\right] cos\frac{n \pi x}{L} $$

Now in this case f(x) is $e^x$ since we are currently dealing with its sine series. So we reduce the expansion of the derivative to...

$$ f'(x) \text{~} \frac{e^L-1}{L} + \sum_{n=1}^\infty \left[ \frac{n \pi}{L}B_n  + \frac2L ((-1)^n e^L - 1)\right] cos\frac{n \pi x}{L} $$

Now, since this f' is the second derivative, we actually have $A_0$! 

$$ A_0 = \frac{e^L-1}{L} $$

Now, of course, we want $A_n$. And while we do have it, this is only by technicality, since it is defined in terms of $B_n$ at the moment. However, we also have $B_n$ in terms of $A_n$ since we agreed the first derivative Was the sine series. So...

$$ A_n = \frac{n \pi}{L}B_n  + \frac2L ((-1)^n e^L - 1) $$
$$ B_n = -\frac{n \pi A_n}{L} $$

Two equations, two unknowns. We should be able to solve for every n. 

$$ A_n = \frac{n \pi}{L}(-\frac{n \pi A_n}{L})  + \frac2L ((-1)^n e^L - 1) $$

$$ \Rightarrow A_n = -\frac{n^2 \pi^2}{L^2}A_n  + \frac2L ((-1)^n e^L - 1) $$
$$ \Rightarrow 1 + \frac{n^2 \pi^2}{L^2} = \frac{1}{A_n}\frac2L ((-1)^n e^L - 1) $$
$$ \Rightarrow \frac{1}{1 + \frac{n^2 \pi^2}{L^2}} = A_n \frac{L}{2((-1)^n e^L - 1)} $$
$$ \Rightarrow \frac{2(-1)^n e^L - 2}{L + \frac{n^2 \pi^2}{L}} = A_n $$

Which, while somewhat ugly and perhaps has a neater form it can be in, certainly functions as a coefficient. The biggest test is the fact that as n gets larger it gets *smaller*. Much smaller. Which is a very good sign.

Since part b) has us explain, we're just going to assume it works and go in blind. So, first of all, when we differentiate both sides, we should note that the derivative of cosine is not sine, it's -sine, so when we differentiate the series we need to watch for the sign change.

First of all, when we differentiate, the $\frac{1}{2a^2}$ goes away, as it is a constant, we are only concerned with terms in the series. So let's look at that...

$$ \frac{2asin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \frac{d}{dx} \left( \frac{(-1)^{n+1}}{n^2-a^2}cos(nx)\right) \right] $$

Now we differentiate. We discussed that the derivative of cosine is -sine, and by the chain rule we pull out another n. The minus sign can combine with the "flipping" -1 term to offset it by one, for raising it to the power of n+2 and n is the same. Thus, we end up with...

$$ = \frac{2asin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \left( \frac{(-1)^{n}n}{n^2-a^2}sin(nx)\right) \right] $$

Now we return to the full equation, with the left side also differentiated.

$$ - a sin(ax) = \frac{2asin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \left( \frac{(-1)^{n}n}{n^2-a^2}sin(nx)\right) \right] $$

$$\Rightarrow sin(ax) = -\frac{2sin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \left( \frac{(-1)^{n}n}{n^2-a^2}sin(nx)\right) \right] $$

Which is what we sought to show. 

The requirements for term-by-term differnetiation with the full Fourier series is if f(x) is piecewise smooth (the trig functions have no discontinuities) and if f(-L) = f(L). In this case, L is $\pi$ and the cosine funciton is even, so this condition is met. Notably we could only do this for the sine function for particular values of a, as it is not even and both sides would have to go to zero. 

The irony, I answered this as a throwaway in the previous psart. sin(ax) is not even, so only certain values of a would allow the second criteria, f(-L) = f(L), to be true. In fact a is specified not to be an integer in the problem, ruining all possible values since L is $\pi$! 

The rule for integration is rather simple: "A Fourier series of piecewise smooth f(x) can always be integrated term by term, and the result is a convergent infinite series that always converges to the integral of f(x)." sin(ax) is definitely piecewise smooth, in fact it is completley continuous in function and derivative form. So we whould just be able to do it. 

So just integrate the series we're given from 0 to x. Curiously, we're not asked to do it from -L to x, but we'll trust the problem. 

$$\int_0^x sin(ax) dx = -\int_0^x \frac{2sin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \left( \frac{(-1)^{n}n}{n^2-a^2}sin(nx)\right) \right] dx $$

$$\Rightarrow [-\frac1a cos(ax)]_0^x = -\frac{2sin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \left( \frac{(-1)^{n}n}{n^2-a^2}\int_0^x sin(nx)dx \right) \right]  $$

$$\Rightarrow -\frac1a cos(ax) + \frac1a = -\frac{2sin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \left( \frac{(-1)^{n}n}{n^2-a^2}(-\frac1n cos(nx) + \frac1n) \right) \right]  $$

$$\Rightarrow cos(ax) = 1+\frac{2asin(a\pi)}{\pi} \left[ \sum_{n=1}^\infty \left( \frac{(-1)^{n}}{n^2-a^2}(- cos(nx) + 1) \right) \right]  $$

So we have the series for cos(ax) in a new form. We can actually pull a clever trick here: $A_0$ is the constant term, so all portions relating to the cos(nx) have to be unrelated to it (or cancel with each other). Therefore..

$$ A_0 = 1+\frac{2asin(a\pi)}{\pi} \sum_{n=1}^\infty \left( \frac{(-1)^{n}}{n^2-a^2} \right)  $$

However, this does not quite complete the problem. It wants us to completley evaluate the infintie series. Naturtally, we would try to compare it to another series. Let's go back and look at the cosine series...



Now the sum we have here is contained in this one, we just need to be a little careful.

$$ \Rightarrow cos(ax) - \frac{2asin(a\pi)}{\pi} \frac{1}{2a^2} = -\frac{2asin(a\pi)}{\pi}\sum_{n=1}^\infty \frac{(-1)^n}{n^2-a^2} cos(nx) $$
$$ \Rightarrow -\frac{\pi}{2asin(a\pi)}cos(ax) + \frac{1}{2a^2} = \sum_{n=1}^\infty \frac{(-1)^n}{n^2-a^2} cos(nx) $$

The problem is now we have this cosine term that is not what we have above. However, we can apply a trick here. We are trying to find $A_0$, and $A_0$ is a constant entirely independent of x. The equality above holds for any x, so what if we just... let x=0 to make our lives easier?

$$ \Rightarrow -\frac{\pi}{2asin(a\pi)} + \frac{1}{2a^2} = \sum_{n=1}^\infty \frac{(-1)^n}{n^2-a^2} $$
$$ \Rightarrow \frac{-\pi a + sin(a\pi)}{2a^2sin(a\pi)} = \sum_{n=1}^\infty \frac{(-1)^n}{n^2-a^2} $$

We can substitute this back into our $A_0$ expression to get:

$$ A_0 = 1+\frac{2asin(a\pi)}{\pi}\frac{-\pi a + sin(a\pi)}{2a^2sin(a\pi)} = 1+\frac{-\pi a + sin(a\pi)}{\pi a} = \frac{sin(a\pi)}{\pi a}$$

Which is exactly what we sought.

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$

So if we need the Fourier sine series, all we need is

$$ f(x) \text{~} \sum_{n=1}^\infty B_n sin\frac{n\pi x}{L} $$
$$ B_n = \frac 2L \int^L_0 f(x) sin\frac{n\pi x}{L} $$

Now with this under our belts, we need to find the coefficients. Notably the integral gets split up, but fortunatley only the middle slice remains as the first and third slices are all "0" and thus integrate to nothing.

$$ B_n = \frac 2L \int_a^{a+\epsilon} \frac \beta\epsilon sin\frac{n\pi x}{L} = \frac 2L \frac \beta\epsilon \left[ -\frac{L}{n\pi}cos\frac{n \pi x}{L}\right]_a^{a+\epsilon} $$


$$ = -\frac{2\beta}{\epsilon n \pi} \left[ cos\frac{n \pi (a+\epsilon)}{L} - cos\frac{n \pi a}{L} \right]$$

Which, while it isn't exactly pretty, does let us define the Fourier sine series. 

$$ f(x) \text{~} -\sum_{n=1}^\infty \frac{2\beta}{\epsilon n \pi} \left[ cos\frac{n \pi (a+\epsilon)}{L} - cos\frac{n \pi a}{L} \right] sin\frac{n\pi x}{L} $$

Now one may think "oh it goes to zero" since the two cosine terms become identical. HOWEVER, we have a discontinuity! There's a $1/\epsilon$ term out front that changes things. Naturally we should turn into a dirac delta function... the quesiton is how exactly. 

The suggestion we received from email was L'hopital's rule. The limit of one funciton divided by another is the limit of the derivatives of both of those functions. Our variable in this case is $\epsilon$. If we treat $\epsilon n \pi$ as the bottom function and everythign else as the topp, we get:

$$ lim_{\epsilon-> 0} f(x) = -\sum_{n=1}^\infty \frac{2\beta}{n\pi}\left[ -\frac{n\pi}{L}sin\frac{n\pi(a+\epsilon)}{L} \right] sin\frac{n\pi x}{L} $$

Via the chain rule for the numerator part. We then apply the limit and cancel terms...

$$ \boxed{lim_{\epsilon-> 0} f(x) \text{~} \sum_{n=1}^\infty \frac{2\beta}{L} sin\frac{n\pi a}{L} sin\frac{n\pi x}{L}} $$

Now this is the way we think the problem was intended to be solved. That said, we did solve it another way first, through a dirac delta function. This is replicated below. 

Perhaps we do not need to prove that the above series is related to the dirac delta function, rather, we just need to take the integral for the sine series *for* the dirac delta function. It is easy to show that the middle function becomes $\beta\delta(x-a)$ at x=a, because we are told the area under the curve (that is, the integral) remains constant, which is essentially the definition of a dirac delta function. Thus the coefficients become...

$$ B_n = \frac 2L \int_0^L \beta \delta(x-a) sin\frac{n\pi x}{L} $$

When the integral happens over a dirac, it pulls out a value of a function, specifically that function at x=a. So we now have:

$$ B_n = \frac 2L \beta  sin\frac{n\pi a}{L} $$

So in this case, we end up with:

$$ f(x) \text{~} \sum_{n=1}^\infty \frac 2L \beta  sin\frac{n\pi a}{L}sin\frac{n \pi x}{L} $$

Which is exactly the same, confirming the result.
