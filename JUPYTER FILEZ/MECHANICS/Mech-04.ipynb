*Consider a double Atwood machine constructed as follows: A mass 4m is suspended from a string that passes over a massless pulley on frictionless bearings. The other end of the string supports a second similar pulley, over which passes a second string supporting a mass of 3m at one end and m at the other. Using two suitable genrealized coordinates, set up the Lagrangian and use the Lagrange equations to find the acceleraiton of the mass 4m when the system is released. Explain why the top pulley rotates even though it carries equal weights on each side.*

PHASE I: 

Write down our energies. Since the pulleys are massless, we just have three masses to worry about. 

$$ \frac12 (m_1v_1^2 + m_2v_2^2 + m_3v_3^2) $$

Kinetic was easy, just adding the energies together. The potential is similar.

$$ g(m_1h_1+m_2h_2+m_3h_3) $$

PHASE II:

Now we need to determine our coordinate system. We have two ropes and thus effectively two degrees of freedom. Let's consider the coordinates to be the location of the second pulley and then the position of the 1m mass with respect to that pulley. Let's call these coordinates $s_1$ for the pulley and $s_2$ for the mass. VERY IMPORTANT: these values will basically always be negative, so watch the signs! 

Because we've defined one of the coordinates on the second pulley, we have to be careful of our coordinate for the first 4m mass. This depends on a quantity not given to us: the length of the string. The string from the second pulley to 4m has a length, say, $l_1$. If the second pulley is at height $- s_1$ then the mass is at height $- l_1 - s_1$. We will keep this length around, however we suspect that it is entirely irrelevant to the acceelration of the system. 

The height of the mass m is just $s_2 + s_1$. This makes our last remaining mass, 3m, have a height of $ - l_2 + s_1 - s_2 $. That gives us our complete potential energy, which is going to see some things nicely cancel.

$$ g(m_1(-l_1 - s_1) + m_2(+ s_2 + s_1) + m_3 (- l_2 + s_1 - s_2))  $$

Placing in the actual values we have for the masses...

$$ = g(4m(-l_1 - s_1) + m(+ s_2 + s_1) + 3m(- l_2 + s_1 - s_2))  $$
$$ = g(-4ml_1 - 4ms_1 +ms_2 + ms_1 -3ml_2 + 3ms_1 -3ms_2)  $$
$$ = g(-4ml_1 -3ml_2 - 2ms_2)  $$

Curiously the potential energy only depends on $s_2$. Not exactly crazy, as with respect to $s_1$ the pulleys are balanced mass-wise. 

We originally thought signs on the velocities would not matter, but the fact is we have multiple velocities adding to produce the final ones, so when we take derivatives we have to be careful.

$$ v_1 = - \dot s_1 $$
$$ v_2 = + \dot s_1 + \dot s_2  $$
$$ v_3 = + \dot s_1 - \dot s_2 $$

We can now put this into the kinetic energy...

$$\frac12 (m_1 \dot s_1^2 + m_2 (\dot s_1^2 + \dot s_2^2 + 2\dot s_1 \dot s_2) + m_3 (\dot s_1^2 + \dot s_2^2 - 2\dot s_1 \dot s_2))$$

We can substitute in the known mass values. And also note that the lengths are completely irrelevant for this part.

$$= \frac12 (4m \dot s_1^2 + m (\dot s_1^2 + \dot s_2^2 + 2\dot s_1 \dot s_2) + 3m (\dot s_1^2 + \dot s_2^2 - 2\dot s_1 \dot s_2))$$
$$= \frac12 (8m \dot s_1^2 + 4m\dot s_2^2 -4m\dot s_1 \dot s_2)$$

To make it very clear what we're working with, here's the model:



PHASE III:

Lagrangian!

$$ \frac12 (8m \dot s_1^2 + 4m\dot s_2^2 -4m\dot s_1 \dot s_2) - g(-4ml_1 -3ml_2 -2ms_2) $$

Just realized that 1/2 can carry through.

$$ 4m \dot s_1^2 + 2m\dot s_2^2 -2m\dot s_1 \dot s_2 + g(4ml_1 +3ml_2 + 2ms_2) $$


PHASE IV:

Apply the EL equation.

$$ D(\partial_2 L) - \partial_1 L = 0 $$

Do it for $s_1$ first.

$$ D(8m\dot s_1 - 2m\dot s_2) - 0 = 0 $$
$$ \Rightarrow 8m\ddot s_1 - 2m\ddot s_2 = 0 $$
$$ \Rightarrow 4\ddot s_1 - \ddot s_2 = 0 $$

This makes the two acceleraitons very closely related to each other, indicating that if one starts moving, the other is going to start moving. There does not exist a situation where, say, masses m and 3m are moving and mass 4m isn't. Which is curious, since the weights are balanced on the first pendulum, but the *motion* of the other weights influence the first one somehow. 

Also note that the lengths are completely irrelevant. As are the masses. 

The accelerations are solved with respect to each other, but we can try to get more information by doing the $s_2$ coordinate.

$$ D(4m\dot s_2 - 2m \dot s_1) - 2mg = 0 $$

Note that the potential energy actually matters here.

$$ \Rightarrow 4m\ddot s_2 - 2m\ddot s_1 - 2mg = 0 $$
$$ \Rightarrow 2\ddot s_2 - \ddot s_1 - g = 0 $$

What we've ended up with is a System of Equations! Whcih means we can get a specific solution, assuming one exists. (quick computer graph confirms that yes, one exists, and with both accelerations being negative to the surprise of no one). 

$$ \ddot s_2 = 4\ddot s_1 $$
$$ 8\ddot s_1 - \ddot s_1 - g = 0 $$
$$ \Rightarrow 7\ddot s_1 = g $$
$$ \Rightarrow \ddot s_1 = g/7 $$

And the other one...

$$ \ddot s_2 = 4g/7 $$

Does this make physical sense? As far as we can tell, yes! $s_2$ directly correlates to the position of the 1m mass, and it is moving up quickly, which is sensible since that second pulley has a 3m on the other side which will definitely push it up. The fact that the other pulley is rolling despite the masses being balanced is odd, but not unexpected given the statement of the problem. 

In specific, since $\ddot s_1 = g/7$, this means the acceleraiton for the 4m mass is $-g/7$, that is it is moving down. Very slowly.  

As for the physical reason why we aren't balanced, that's because there's motion in the system. if m and 3m were staitonary, nothing would happen, but 3m pulls more than m and thus drags m up. This creates an offset--m is moving up, 3m is moving down, and the acceleraiton alters the tension in the rope, which in turn influences how much their junction pulls on the first rope. If the two masses were equal, one could argue that the offsets to the tension would cancel out, but here they are *not* equal, there's more change coming from 3m than m, thus the tension is not simply just the force of 4m and is in fact tipped slightly in the favor of letting the 4m mass "win out." 

One could also argue this from conservation of angular momentum. m and 3m could be thought of as rotating "around" each other as they move, and to offset that angular momentum the 4m mass would have to move. 

Fortunately we were not asked to find this in the Newtonian method...

*Derive the Lagrange equations for the restricted three-body problem, given the Lagrangian 1.150. Identify the Coriolis and centrifugal force terms in your equations of motion.*

1.150: $L = \frac12 m (\dot x_r^2 + \dot y_r^2) + \frac12 m \Omega^2 (x_r^2 + y_r^2) + m\Omega(x_r\dot y_r - \dot x_r y_r) + \frac{GM_0m}{r_0} + \frac{GM_1m}{r_1}$

So this is a little odd, we can't go through our steps since we are already presented with the answer for PHASE III, the Lagrangian. However, to even begin PHASE IV, we need to understand our system. What coordinates even are there here?

Well, this is the *restricted* three body problem, the larger masses are locked on a circular orbit, it's only the smaller mass that can act in unpredictable ways. That mass has two coordinates in this Lagrangian: $x_r$ and $y_r$, so that is what we will need to base our answers off of.

We need to be careful though, since the radii are not just constants.

$$ r_0 = \sqrt{(x_r+a_0)^2 + y_r^2} $$
$$ r_1 = \sqrt{(x_r-a_1)^2) + y_r^2} $$

So we're going to have some funky derivatives at the very least. Fortunately this isn't a math class and we won't have to yell at said derivatives to get them to behave.

Anyway, we seek the lagrange equations. If it is convenient we will also grab the full equations of motion, but we rather doubt that will be the case. 

Anyway, E-L equations are $D(\partial_2 L) - \partial_1 L = 0$ as usual. Start with x.

$$ D(2m\dot x_r - m\Omega y_r) - m\Omega^2 x_r - m\Omega \dot y_r - \frac{GM_0m(-x_r-a_0)}{r_0^3} - \frac{GM_1m(-x_r+a_1)}{r_1^3} = 0 $$

$$ \Rightarrow 2\ddot x_r - \Omega \dot y_r - \Omega^2 x_r - \Omega \dot y_r - \frac{GM_0(-x_r-a_0)}{r_0^3} - \frac{GM_1(-x_r+a_1)}{r_1^3} = 0 $$

Which is the Lagrange equation for motion in x. Quite a mess, obviously, but at least the semgnets are easily noticable.

The version for y is very similar as the system is more or less symmetric.

$$ D(2m\dot y_r + m\Omega x_r) - m\Omega^2 y_r + m\Omega \dot x_r + \frac{GM_0m(y_r)}{r_0^3} + \frac{GM_1m(y_r)}{r_1^3} = 0 $$

$$ \Rightarrow 2\ddot y_r + \Omega \dot x_r - \Omega^2 y_r + \Omega \dot x_r + \frac{GM_0(y_r)}{r_0^3} + \frac{GM_1(y_r)}{r_1^3} = 0 $$

As said, very similar, but with notable sign differences. 

Now, though, we aren't done. We still need to identify the coriolis and centrifugal forces! 

In order to do that, we can get rid of the forces that are clearly the *gravitational* forces-- the ones with big G in them, and only examine what remains.

$$ 2\ddot x_r - \Omega \dot y_r - \Omega^2 x_r - \Omega \dot y_r = (gravity) $$
$$ 2\ddot y_r + \Omega \dot x_r - \Omega^2 y_r + \Omega \dot x_r = (gravity) $$

Oh hey, like terms we didn't notice.

$$ 2\ddot x_r - 2\Omega \dot y_r - \Omega^2 x_r = (gravity) $$
$$ 2\ddot y_r + 2\Omega \dot x_r - \Omega^2 y_r = (gravity) $$

With this, we can clearly identify two kinds of forces: $\Omega^2$ and $\Omega$. The question is which is which? Well, the identifiers we used to label them earlier tell us: $\Omega$ versus $\Omega^2$. These represent angular velocoties! A quick look up of the definitions of Coriolis Force and Centrifugal Force reveal them to be $-2m\Omega v$ and $m \Omega^2 r$ respectively. 

While our versions of the forces are split up in components, we can see them in exactly the right forms. Coriolis is from the $\Omega$ terms and centrifugal is from the $\Omega^2$ terms. They even have the correct dependence on eitehr velocity or position. Tah-dah!
