*Explore the dynamics of the driven pendulum, using the surface of section method. We are interested in exploring the regions of a parameter space over which various phenomena occur. Consider a pendulum of length 9.8m, mass 1kg, and acceleration of gravity g = 9.8, giving $\omega_0$ = 1. Explore the parameter plane of the amplitude A and frequency $\omega$ of the periodic drive. Examples of the phenomena to be investigated:*

*a) Inverted equilibrium. Show the region of parameter space (A,$\omega$), in which the inverted eqilibrium is stable. If the inverted equilibrium is stable there is some range of stability, i.e., there is a maximum angle of displacement from the eqiulibrium that stable oscillations reach. If you have enough time, plot contours in the paramater space for different amplitudes of the stable region.*

So, for this problem we are asked to explore the parameter space for $(A,\omega)$ and bound certain behaviors. To do this we might as well just start searching out a grid of integer values and figure out what lies where. Since this will be a lot of different runs, we won't reproduce most of them here. We will start at (0,0) and cycle outward, including negative values until it's clear there's some kind of mirroring. (0,0) produces nothing, as does anything with $\omega=0$. (0,1) and (0,-1) are completely regular. (1,1), (1,-1), (-1,1) and (-1,-1) all produce the same graph, so we determine that we can restrict A and $\omega$ to positive values and still get what we need. The shape they all produce is a single island of stability "eye" surrounded by chaos, and then at the higher momentums are circulation paths. This notably does not match what we are looking for, yet.

Next, we try (1,2) (2,1) and (2,2). (1,2) and (2,1) are *not* symmetric. (1,2) produces a similarly shaped eye graph to what we got previously. (2,1) produces an hourglass plot with two outer islands and circulating paths. (2,2) has no central island, but it has several smaller islands spread throughout a "pinched" area of chaos. 

Now that we have our methods outlined we will stop dictating what exactly we see in every single plot, as that won't exactly be helpful. For part a), we're looking for areas where we have inverted equilibrium. These exist as the stable regions above and below the central one arranged at $\pi$ and -$\pi$. From experiments, we find that these never exist at A=0 or $\omega=1$. We will also only consider the apparently contiguous region that includes A=1 $\omega=2$ as that's the lowest section where we see it.

The problem is the region seems to have no hard borders. In the vast, vast, vast majority of sections, there exist the inverted equilibrium points. The central equilibrium is gone a lot more than the inverted ones! 

For instance, along A=1, it does not matter how large we increase $\omega$, the inverted equlibrium *always* exist, even when we reach *thousands* of $\omega$. We were unable to find any value of $\omega$ that provided a boundary--there were always inverted equilibrium poits, and even increasing it to absurd levels did not remove the region of chaos between them.

However, increasing A does seem to bound it, but we are unable to determine exactly when or how. Considering each value of $\omega$ separately, we found that $\omega=2$ has inverted equilibrium at A=3 and no further, 3 has 8, and 4 had 14. Thus we can imagine a sort of triangular section cutting across the space, IF we consider each line in isolation. However, this does not work. As an illustration, consider the matrix, where O represents a situation where the inverted equilibrium exists, and X where it doesnt.

$$\begin{bmatrix}
O & O & O & O & X\\
O & X & X & O & O\\
\end{bmatrix}$$


Even though each individual row has a definite "end", the fact is we also have to consider *other adjacent points* on this sampling, which means that even though the bottom row ended at the second column, the continguous region of O continued past it. So we can't really say for sure what the bound is by inspection.

We considered the fact that we might be undersampling and that there might be more defined limits at fracitonal $\omega$, but this does not appear to be the case.

So we tentatively say that "well, these regions cannot exist at $\omega < 2$ and $A < 1$. There is absolutely no limit on $\omega$, and there appears to be a limit for large A but it's not easily predictable. 

*b) Period doubling of the normal equilibrium. For this case, plot the angular momenta of the stable and unstable equilibria as functions of the frequency for some given amplitude*

From inspections in part a) the situation where A=3 is actually the most interesting. Some of the early $\omega$s have no stable or unstable equilibrium, but this changes as time goes on. We measure the momentum by eyeballing the location of the centers of the islands of stability. The stable equilibrium will generally have an angular momentum of zero, but not always, it does split into two sometimes. 

Anyway, after examining $\omega$ 1 through 10, we find that the stable equilibrium was always centered at 0 momentum save for once at $\omega=3$, where it split into two, each at 100 angular momentum. 

Below is the graph for the unstable equilibria. 



As we can see the momentum in the unstable equilibria increases gradually with the frequency, though there appears to be a slight tapering effect (but we aren't sure, that could just be due to inaccurate measurements of the centers of the graphs). 

*c) Transition to large-scale chaos. Show the region of parameter space (A,$\omega$) for which the chaotic zones around the three eprincipal resonance islands are linked.*

Now this we might be able to parameterize easier, as it requires all *three* equilibria to exist *and* for the chaotic regions to all be linked. For A=1 this occurs rather rigidly at $\omega=7$, when the central stability island reaches the edges of the graph, splitting the chaos in two. Also, $\omega=1$ never has any unstable equilibria, and A=5 appears to be the cutoff ofr the central stability existing. 

Exploring further, we are able to make this graph of the regions where all three exist and the chaotic zones are linked. We consider poitns where the central equilibrium is an hourglass or connected regions to be acceptable. 

We discover that A=5 is a sharp cutoff, after that no central equilibrium exists even at absurd frequencies. For A=1, we are cut off at a frequency of 7, and for A=2 we are cut off somewhere around 28 (it's a little hard to tell when the chaotic zone finally condenses into a line.). For A=3 and A=4 there appears to be no limit to how large the frequency can get and all three stable zones still exist. 

As for lower bounds, for A=1,2, they are $\omega=2$, and for A=3,4 it is $\omega=3$. 

We were going to make an image but with 3 and 4 going off to infinity, and 2 going to 28, it'd just be a very tall tower. Not the best use of the time.
