# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

First, solve the Lagrangian Problem.

PHASE I: Energies?

Two kinetic, $\frac12m(v_1^2 + v_2^2)$

Three potential $T(d_1+d_2+d_3)$

No gravitaitonal potential energy, the table supports us and is frictionless. 

It is conceivable that each rope could have its own tension, but we are told they are the same, and that they all have the same lengh as well. 

PHASE II: Coordinates

Since everything is only moving vertically, the only coordinates we need are the vertical displacements of the two masses. Let's call this $y_1$ and $y_2$, as is pretty heavily implied by the problem. Also, since $y_1 << l >> y_2$ this *also* means the small angle approximation holds. so if we ever seen an opportunity to use that, we should.

Anyway, the velocities are just $\dot y_1, \dot y_2$, nothing special. 

The potentials are found by the Pythagorean theorem.

$$ d_1 = \sqrt{l^2 + y_1^2} $$
$$ d_2 = \sqrt{l^2 + (y_1-y_2)^2} $$
$$ d_3 = \sqrt{l^2 + y_2^2} $$

The order of the subtraciton doesn't matter, as the quantity is squared. 

PHASE III: Lagrangian

We then construct the Lagrangian, T-U.

$$ \frac12 m(\dot y_1^2 + \dot y_2^2) - T(\sqrt{l^2+y_1^2} + \sqrt{l^2+(y_1-y_2)^2} + \sqrt{l^2+y_2^2}) $$

There sadly doesn't seem to be a way to simplify this, at least not right away. 

PHASE IV: The E-L Equation

$$ D(d_2 L) - d_1 L = 0 $$

Apply this to both coordinates to arrive at...

$$ m\ddot y_1 = -T \left( \frac{y_1-y_2}{\sqrt{l^2+(y_1-y_2)^2}} + \frac{y_1}{\sqrt{l^2 + y_1^2}} \right) $$
$$ m\ddot y_2 = -T \left( \frac{y_2-y_1}{\sqrt{l^2+(y_1-y_2)^2}} + \frac{y_2}{\sqrt{l^2 + y_2^2}} \right)$$

Yes we just had a computer take this derivative.

Now we need to switch gears into solving the Normal Modes problem. We have already done PHASE I for this problem--solve the Lagrangian problem. Next is...

PHASE II: Matrices

$$ \textbf{M} \vec q'' = -\textbf{K} \vec q $$

(q is primed instead of dotted since LATEX throws a fit about vectoring and dotting a letter at the same time). 

We have to find out what these matrices are. In both cases they will be 2x2 since we have two variables we are concerned with. The mass matrix is trivial.

$$ \textbf{M} = \begin{bmatrix} 
m & 0 \\
0 & m
\end{bmatrix}$$

The K matrix, though, is much more problematic as we are not simply able to split everything up into terms of each veriable... unless we apply the approximation! If we do that, all the roots simply become l as l >> y in both cases. Then the K matrix can be found somewhat easily.

$$ \textbf{K} = \frac{T}{l} \begin{bmatrix} 
2 & -1 \\
-1 & 2
\end{bmatrix}$$

PHASE III: The Characteristic Equation

$$ det(\textbf K - \omega^2 \textbf M) = 0 $$

Once we plug this in we get...

$$ \left|\begin{matrix} 
2\frac{T}{l} - \omega^2 m  & -\frac{T}{l} \\
-\frac{T}{l}  & 2\frac{T}{l} - \omega^2 m
\end{matrix}\right| = 0$$
$$ \Rightarrow = (4\frac{T^2}{l^2} + \omega^4 m^2 - 4 \omega^2 m \frac{T}{l}) - (\frac{T^2}{l^2}) = 0 $$
$$ \Rightarrow = 3\frac{T^2}{l^2} + \omega^4 m^2 - 4 \omega^2 m \frac{T}{l} = 0 $$
 
PHASE IV: Solve.  

Now we can solve this as a quadratic equation with $\omega^2$ as the variable. Since we are a little lazy we just have the computer do this.

$$ \omega^2 = \frac{3T}{lm}, \frac{T}{lm} $$
$$ \Rightarrow \omega = \sqrt{\frac{3T}{lm}}, \sqrt{\frac{T}{lm}} $$

The factor of 3 is perhaps unsurprising, since it showed up in normal mode examples preivously. 

PHASE V: magnitudes. 

Now one thing we did not mention is that the Characteristic Equation is actually a form of the Eigenvalue Equation

$$ (\textbf K - \omega^2 \textbf M) \vec a = 0 $$
$$ a = Ae^{i\delta} $$

Where the exact expression is different depending on what normal mode we use. The important values to note are the amplitudes A and the phases $\delta$, which will tell us how the various particles move with respect to each other. We predict one set that are equal in amplitude and phase, and one set that have equal amplitude but opposite phase. 

Anyway, to solve these, we need to examine the combined matrix. Let's start with the simpler normal mode first.

$$ \begin{bmatrix} 
\frac{T}{l} & -\frac{T}{l} \\
-\frac{T}{l}  & \frac{T}{l}
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2}
\end{bmatrix}= 0$$

Which means...

$$ A_1 = A_2 $$
$$ \delta_1 = \delta_2 $$

Which is equivalent to motion where the two masses are moving up and down in unison, as expected.

Now if we do the other normal mode...

$$ \begin{bmatrix} 
-\frac{T}{l} & -\frac{T}{l} \\
-\frac{T}{l}  & -\frac{T}{l}
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2}
\end{bmatrix}= 0$$

This means...

$$ A_1 = -A_2 $$
$$ \delta_1 = \delta_2$$

Wich is equivalent to having the amplitudes equal and the phases exactly 180 degrees out of phase. Which corresponds to the normal mode where one is going up and one is going down, as expected.

We note that the problem only asks for us to find the normal modes and describe their motion, so we do not seek to find what A actually IS.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

As always... we start with solving the Lagrangian. Whose first step is the first thing we're told to find!

PHASE I: Energies.

Kinetic: $\frac12 m (v_1^2 + v_2^2)$

Potential: $mg(h_1+h_2) + \frac12 kx^2$ 

Where x is the extension of the spring.

PHASE II: Coordinates.

The coordinates are provided for us, actually, two angles, $\varphi_1, \varphi_2$. 

Since we are explicilty using the small angle approximation, the spring extension is approximated by $l(\varphi_2 - \varphi_1)$, as told in the problem. 

The velocities are attached to just one angle, specifically $l\dot\varphi$, one for each mass.

Heights are just $-lcos\varphi$.

PHASE III: Lagrangian.

$$ \frac12 m(l^2\dot\varphi_1^2 + l^2\dot\varphi_2^2) + mgl(cos\varphi_1 + cos\varphi_2) - \frac12 k l^2 (\varphi_2-\varphi_1)^2$$

PHASE IV: E-L Equations.

We have two coordinates so we use $D(\partial_2 L) - \partial_1 L = 0$ twice.

$$ D(ml^2\dot\varphi_1) + mglsin\varphi_1 - kl^2(\varphi_2 - \varphi_1) = 0  $$
$$ ml^2\ddot\varphi_1 + mglsin\varphi_1 - kl^2(\varphi_2 - \varphi_1) = 0  $$
$$ \ddot\varphi_1 = \frac{-mglsin\varphi_1 + kl^2(\varphi_2 - \varphi_1)}{ml^2} $$
$$ \ddot\varphi_1 = \frac{-mgsin\varphi_1 + kl(\varphi_2 - \varphi_1)}{ml} $$

Which does contain the pendulum term, as it should. 

The other term is almost identical, the only difference being a sign on the spring term.

$$ \ddot\varphi_2 = \frac{-mgsin\varphi_2 - kl(\varphi_2 - \varphi_1)}{ml} $$


With the Lagrange Equations of Motion down, we move on to the normal mode problem. First phase is already done, our lagrangian is complete. Now we deal with...

PHASE II: Matrices

$$ \textbf{M} \vec q'' = -\textbf{K} \vec q $$

The mass matrix will have us move our mass back from the denominator to the left hand side, creating an expected matrix:

$$ \textbf{M} = \begin{bmatrix} 
m & 0 \\
0 & m
\end{bmatrix}$$

The K matrix, though, will have a few more terms, and terms that are not easily reducable to ismple numbers. Note that in our small angle approximations the sines of the angles just become the angles themselves.

$$ \textbf{K} = \begin{bmatrix} 
\frac{mg}{l}+k & -k \\
-k & \frac{mg}{l}+k
\end{bmatrix}$$

Sensibly symmetric.

PHASE III: The Characteristic Equation

$$ det(\textbf K - \omega^2 \textbf M) = 0 $$

Once we plug this in we get...

$$ \left|\begin{matrix} 
\frac{mg}{l}+k-\omega^2m & -k \\
-k & \frac{mg}{l}+k-\omega^2m
\end{matrix}\right| = 0$$
$$ \Rightarrow = (\frac{m^2g^2}{l^2} + k^2 + 2\frac{mgk}{l} + \omega^4m^2 - 2\frac{mg}{l}\omega^2m - 2k\omega^2m ) - (k^2) = 0 $$
$$ \Rightarrow = \frac{m^2g^2}{l^2} + 2\frac{mgk}{l} + \omega^4m^2 - 2\frac{mg}{l}\omega^2m - 2k\omega^2m = 0 $$
$$ \Rightarrow = \frac{m^2g^2}{l^2} + 2\frac{mgk}{l} + \omega^4m^2 - \omega^2 \left(2\frac{m^2g}{l} + 2km \right) = 0 $$

Which can be solved as a quadratic with $\omega^2 = x$. 

 
PHASE IV: Solve.  

The quadratic formula provides...

$$ \omega^2 = \frac{g}{l} + \frac{k}{m} \pm \frac{k}{m} $$

Which is to say g/l or g/l + 2k/m. 

Technically speaking, these are the normal modes.

$$ \omega = \sqrt{g/l + k/m \pm k/m} $$

Now we just need to explain what these normal modes are, the problem doesn't ask us to do more than find them. The case where $\omega= \sqrt{g/l}$ is very obviously just the situation where the spring isn't compressing at all and the entire system acts as a single pendulum. However, the other case...

$$ \omega = \sqrt{g/l + 2k/m} $$

is not so clear what it represents, though we can guess that the pendulums are completely anti-alligned. So for this case we go into

PHASE V: Magnitudes

$$ (\textbf K - \omega^2 \textbf M) \vec a = 0 $$
$$ a = Ae^{i\delta} $$

The combined matrix is...

$$ \begin{bmatrix} 
\frac{mg}{l}+k-\omega^2m & -k \\
-k & \frac{mg}{l}+k-\omega^2m
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2}
\end{bmatrix}= 0$$

With the specific frequency we care about this becomes

$$ \begin{bmatrix} 
-k & -k \\
-k & -k
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2}
\end{bmatrix}= 0$$

This pattern should be familiar to that of the matrix in the previous problem, note how the matrix is all single values that have the same sign, it's a constant matrix. This means the only solutions are if the amplitudes are opposite or the phases are 180 degrees separate. Either of these is equivalent to the motion we predicted: two pendulums moving exactly opposite each other stretching and squishing the spring in the middle. 

With that, we have found and described both normal modes.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

We could just eyeball the normal coordinates based on our previous solution, but there is an actual procedure for this.

$$ \vec x = x_1 \vec A_1 + A_2 \vec a_2 = (x_1 + x_2 , x_1 - x_2)$$

Except in our case this would be the angles.

$$(\varphi_1 + \varphi_2 , \varphi_1 - \varphi_2)$$

Now, to see if we interpreted this correctly, we would have to show that they are decoupled. Which means running through the Lagrangian again. Also, while scalars out front shouldn't matter too much, we slap a 1/2 out front for scaling. (We believe so everything can be added together to produce the original coordinates?)

$$ \vec x = \frac12(\varphi_1 + \varphi_2 , \varphi_1 - \varphi_2) = \vec\zeta $$

Now that we know the coordinates we want to use, start from the beginning again. We already have the energies, but these new coordinates will redefine how the Lagrangian appears. So we must rewrite the energies entirely. Fortunately we can put them in terms of our old coordinates:

$$\zeta_1 + \zeta_2 = \varphi_1$$
$$\zeta_1 - \zeta_2 = \varphi_2$$

This also works for the velocities.

$$\dot\zeta_1 + \dot\zeta_2 = \dot\varphi_1$$
$$\dot\zeta_1 - \dot\zeta_2 = \dot\varphi_2$$

So one can take the Lagrangian

$$ \frac12 m(l^2\dot\varphi_1^2 + l^2\dot\varphi_2^2) + mgl(cos\varphi_1 + cos\varphi_2) - \frac12 k l^2 (\varphi_2-\varphi_1)^2$$

and convert it to:

$$ ml^2(\dot\zeta_1^2 + \dot\zeta_2^2) + 2mgl(cos\zeta_1 cos\zeta_2) + k l^2 (\zeta_2)^2$$

One may argue at this point we have a clearly coupled term, the two cosines right there have one with each variable, and are coupled. But wait, remember, once we get to the end we are in the small angle approximation. When we take the derivative of that section, we will end ot with a $sin\zeta_1cos\zeta_2$ pattern (or the reverse). In the small angle the sine equals the angle, and will dominate over the cosine term which can just be represented as 1. 

Thus, these really are normal coordiantes, in the small angle approximation. 

Now, though, we apply a force to this situation. We would like to just do the Lagrangian again and solve, proving that everything is uncoupled, but this force isn't a constraint so we can't use Lagrange Multipliers, so we have to come up with some other way to reason this out. 

To do this, we shall put things in the form F=ma, which means actually doing the E-L equatins in full on the modified Lagrangian.

$$ ml^2\ddot\zeta_1 + 2mgl\zeta_1 = 0 $$
$$ ml^2\ddot\zeta_2 + 2mgl\zeta_2 + 2kl^2\zeta_2 = 0 $$

Which become...

$$ m\ddot\zeta_1  = -m\frac{g}{l} \zeta_1 $$
$$ m\ddot\zeta_2 = -m\frac{g}{l} \zeta_2 - 2k\zeta_2$$

So these are definitely uncoupled. We can relate them to the forces on the individual bobs via:

$$ \frac12(\ddot\varphi_1 + \ddot\varphi_2 , \ddot\varphi_1 - \ddot\varphi_2) =(\ddot \zeta_1 , \ddot \zeta_2)$$

Let's take the first one as an example since both will be nearly identical.

$$ \ddot\zeta_1 = \frac12 (\ddot\varphi_1 + \ddot\varphi_2) $$
$$ \Rightarrow m\ddot\zeta_1 = \frac12 (m\ddot\varphi_1 + m\ddot\varphi_2) $$

If we apply the extra force, bv, it will change the ma terms for their respective values.

$$ m\ddot\zeta_1 - F = \frac12 (m\ddot\varphi_1 - bl\dot\varphi_1 + m\ddot\varphi_2 - bl\dot\varphi_2) $$
$$ \Rightarrow m\ddot\zeta_1 - F = \frac12 (m\ddot\zeta_1 - bl\dot\zeta_1) $$

Thus the mysteirous force on $\zeta_1$, F, is dependent only on the velocity of $\zeta_1$. Thus, we are still independent of $\zeta_2$. The relation is virtually identical for the other coordinate. 

$$ m\ddot\zeta_2 - F = \frac12 (m\ddot\varphi_2 - bl\dot\varphi_2 - m\ddot\varphi_2 + bl\dot\varphi_2) $$
$$ \Rightarrow m\ddot\zeta_2 - F = \frac12 (m\ddot\zeta_2 - bl\dot\zeta_2) $$

Notably this actually didn't require the full F=ma equations like we thought... oh well, it's good to have them anyway. 

Anyway, now we need to find the motion with respect to our new coordinates. This is going to be a bit more awkward, since our chosen coordinates are slightly... odd, but we can still do it.

Good thing we bothered to find the final form of the Lagrange Equations.

$$ m\ddot\zeta_1  = -m\frac{g}{l} \zeta_1 $$
$$ m\ddot\zeta_2 = -m\frac{g}{l} \zeta_2 - 2k\zeta_2$$

Now the modes better be the same, so no need to find $\omega$ again, we have

$$ \omega = \sqrt{g/l + 2k/m}, \sqrt{g/l} $$

However we still need to create our matrices. Fortunately both of them are just diagonal matrices. Which is the entire poitn of normal coordinates. 

$$ \textbf{M} = \begin{bmatrix} 
m & 0 \\
0 & m
\end{bmatrix}$$

$$ \textbf{K} = \begin{bmatrix} 
mg/l & 0 \\
0 & mg/l+2k
\end{bmatrix}$$

The full matrix eigenvalue equation is now...

$$ \begin{bmatrix} 
mg/l - \omega^2m & 0 \\
0 & mg/l - k - \omega^2m
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2}
\end{bmatrix}= 0$$

For the basic pendulum motion this becomes

$$ \begin{bmatrix} 
0 & 0 \\
0 & 2k
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2}
\end{bmatrix}= 0$$

And for the pendulums at inverse phase...

$$ \begin{bmatrix} 
2k & 0 \\
0 & 0
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2}
\end{bmatrix}= 0$$

This basically says that in mode 1 only the first coordinate is oscillating, and in mode 2 only the second coordinate is oscillating. This is because for 2kA = 0 to be true, the amplitude A of that relation has to be zero, leaving only the other portion free to move. 

So yes, we have completley decoupled them. As for describing the motion, it's the exact same as before: both pendulums moving in unison, or presicely 180 degrees out of phase.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

So, let us begin... with the Lagrangian problem. Let's label the objects 1 to 3 from left to right. 

PHASE I: Energies.

$$\frac12 m (v_1^2 + v_3^2) + \frac12 M v_2^2$$

$$\frac12 k (x_1^2 + x_2^2)$$

We have three objects and two springs, it makes sense.

PHASE II: Coordinates

The picture itself suggests that we just use three coordiantes, one for each mass. One could try to argue that we could set the big mass M as the origin, but let's go ahead and go with the suggested coordiantes. (The problem implies that one of the frequencies will be zero, which we think is the mode the "setting M as origin" would have gotten rid of). 

Regardless, this means our spring displacements are $x_2 - x_1$ and $x_3 - x_2$ respectively. The velocities are just the time derivatives of the three cooridnates.

PHASE III: Lagrangian

$$ \frac12m(\dot x_1^2 + \dot x_3^2) + \frac12 M\dot x_2^2 - \frac12 k (2x_2^2 + x_1^2 + x_3^2 - 2x_1x_2 -2x_2x_3 ) $$

PHASE IV: E-L Equation.

Do it once for each coordiante, but there shouldnt' be anything unusual here.

$$ m\ddot x_1 + kx_1 -kx_2 = 0$$
$$ M\ddot x_2 + 2kx_2 -kx_1 - kx_3 = 0$$
$$ m\ddot x_3 + kx_3 -kx_2 = 0$$

Now we shift to the normal mode problem. The Lagrangian is the first part, the second part is

PHASE II: Matrices.

$$ \textbf{M} = \begin{bmatrix} 
m & 0 & 0 \\
0 & M & 0 \\
0 & 0 & m
\end{bmatrix}$$

$$ \textbf{K} = \begin{bmatrix} 
k & -k & 0 \\
-k & 2k & -k \\
0 & -k & k
\end{bmatrix}$$

PHASE III: The Characteristic Equation

$$ \left|\begin{matrix} 
k - \omega^2 m & -k & 0 \\
-k & 2k - \omega^2 M & -k \\
0 & -k & k - \omega^2 m
\end{matrix}\right| = 0$$

We just had a computer take this determinant. 

$$ -2k^2m \omega^2 - k^2M \omega^2 + 2km^2 \omega^4 + 2kmM \omega^4 - m^2M \omega^6 = 0 $$

PHASE IV: Solve
    
We can clearly see that $\omega=0$ is a solution here, as it shoudl be. So let's divide it out until we get a constant term.

$$ \Rightarrow -2k^2m - k^2M + 2km^2 \omega^2 + 2kmM \omega^2 - m^2M \omega^4 = 0 $$
$$ \Rightarrow -k^2(2m + M) + 2 \omega^2 k m (m + M) - m^2M \omega^4 = 0 $$

Which we can solve with the quadratic equation like the previous problems.



$$\omega^2 = \frac{km + kM \pm km}{mM} = \frac{2km + kM}{mM} , \frac{k}{m}$$

These are two nonzero frequencies. The third one *is* zero. 

With all our tools we can now examine the eigenvalue equaiton.

$$ \begin{bmatrix} 
k - \omega^2 m & -k & 0 \\
-k & 2k - \omega^2 M & -k \\
0 & -k & k - \omega^2 m
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2} \\
A_3e^{-i\delta_3}
\end{bmatrix}= 0$$


Let's begin with the 0 case first as it's the simplest.

$$ \begin{bmatrix} 
k & -k & 0 \\
-k & 2k  & -k \\
0 & -k & k 
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2} \\
A_3e^{-i\delta_3}
\end{bmatrix}= 0$$

Which gives us, if we look at only the amplitudes...

$$ A_1 - A_2 = 0 $$
$$ -A_1 + 2A_2 - A_3 = 0 $$
$$ A_3 - A_2 = 0 $$

Which means that $$A_1 = A_2 = A_3$$. However, all three oscillating in synch doesn't make sense as there is nothing to pull them back and forth if they're all moving at the same amplitude... so this corresponds to a *non-oscillatory* result. Which, duh, $\omega=0$. Naturally the amplitudes, or positions, will all be the same if the molecule as a whole were drifting in one direction without being perturbed.

And this confirms the earlier theory that if we defined the coordinate system with M as the center, that we would have had no zero result.

Let us hnow consider k/m. 

$$ \begin{bmatrix} 
0 & -k & 0 \\
-k & 2k - kM/m & -k \\
0 & -k & 0 
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2} \\
A_3e^{-i\delta_3}
\end{bmatrix}= 0$$


that awkward central term would be really annoying, if the other two equations didn't immediately demand that $A_2=0$. So what we get out of this is

$$ -A_1 -A_3 = 0 $$

Which is to say the rather interesting result of $A_1 = -A_3$, that is, the outer masses are moving out at the same time or in at the same time, making the entire setup pulse back and forth. 

Lastly, (2km + kM)/(mM)

$$ \begin{bmatrix} 
- 2km/M & -k & 0 \\
-k & -kM/m & -k \\
0 & -k & - 2km/M
\end{bmatrix}\begin{bmatrix}
A_1e^{-i\delta_1}\\
A_2e^{-i\delta_2} \\
A_3e^{-i\delta_3}
\end{bmatrix}= 0$$

This clearly results in something a lot less *normal*. From this we can get the following

$$ (2m/M)A_1 + A_2 = 0 $$
$$ A_1 + (kM/m)A_2 + A_3 = 0 $$
$$ (2m/M)A_3 + A_2 = 0 $$

From this we quiclky see that $A_1 = A_3$ and that $A_2$ has to have the inverse sign of the other two. This means that while the outer weights are going one way, the inner weight is going the other, and they trade off in this back and forth motion. 

So, the three motions are: translational, in-out vibration, and inner-outer vibration, if trying to be concise. 

So, for the last part, we are finding the k value for CO2. The values we are given correspond to our frequencies.

$$\omega = \sqrt{\frac{2km + kM}{mM}}, \sqrt{\frac{k}{m}}$$

The values we are given are 2565$cm^{-1}$ 1480$cm^{-1}$. Converting to inverse meters, we get 2.565e5 and 1.480e5, respectively. We do have to ask which frequency corresponds to the ones we found, but our first result is clearly larger than the second, so that answers our question. In order to make this easier to work with we use the squared values of these quantities when evaluating. 

The mass of carbon (M) is 1.99e-26 kg, and oxygen (m) is 2.657e-26 kg. With this, we have all we need to solve for k. We'll go ahead and solve it twice, the results should be comparable for both frequencies. 

$$ k = \omega_1^2 m = (5.819e-16) kg/m^2 $$
$$ k = \omega_2^2 \frac{mM}{2m+M} = (4.763e-16) kg/m^2 $$

We note that these don't agree...
