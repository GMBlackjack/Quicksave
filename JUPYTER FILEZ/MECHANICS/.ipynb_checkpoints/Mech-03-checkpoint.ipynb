*A mead of mass m is constrained to move on a frictionless helical wire. The helix is oriented so that its axis is horizontal. The diameter of the helix is d and its pitch (turns per unit length) is h. The system is in a uniform gravitaitonal field with vertical acceleration g. Formulate a Lagrangian that describes the system and find the Lagrange equations of motion*

PHASE I

The kinetic energy of this system is the usual $\frac12mv^2$

The potential energy can be written as mgh. 

PHASE II

The system has one degree of freedom, along the wire. Thus we choose a single coordinate, call it s. s is just the length along the wire, and the velocity rather simply follows from this. 

Does this change anything about our energies from phase I?

Well, the velocity is still the same, but we can write it as $Ds$ now, so the kinetic energy is $\frac12m(Ds)^2$.

The height in the potential energy is a bit different though, for that h is vertical height, not helical distance. We just need to construct the value height from the given d, h, and s. Please note that the h's are different, height and pitch are not the same thing. Thus when we refer to the "height" we shall now call it y, to avoid problems.

So, we need y(s). This is going to involve knowing how long a helical strand *is* after it's gone all the way around once. If we imagine unfolding a single coil of a helix, we'll get a wire that forms the hypotenuse of a right triangle. One side is the circumference of the sircle, $d\pi$. The other side is the height it takes to complete one turn. The value we are given is in turns per unit length, since we want length per turn, we find the value of this to be 1/h. 

So, for a single turn, 

$$ s = \sqrt{d^2\pi^2 + \frac{1}{h^2}} $$

Now, we want to be able to do this for any height at all. Well, since we know the height for one loop is 1/h, we can scale it. Half a loop would have 1/2h, while it would have $d\pi/2$ as well. Noting this scale, we can rewrite...

$$ s = \sqrt{h^2y^2d^2\pi^2 + y^2} $$

Now we want to solve for y here...

$$ \Rightarrow s = y \sqrt{h^2d^2\pi^2 + 1} $$
$$ \Rightarrow \frac{s}{\sqrt{h^2d^2\pi^2 + 1}} = y $$

This'll do. This means our potential energy is now 

$$ mg\frac{s}{\sqrt{h^2d^2\pi^2 + 1}} $$

PHASE III

Now we find the Lagrangian. This is just kinetic minus potential energy. So...

$$ L = \frac12m(Ds)^2 - mg\frac{s}{\sqrt{h^2d^2\pi^2 + 1}} $$

Nothing else to do in this phase.

PHASE IV

With the Lagrangian in tow, now we turn to the E-L equation. We only have one coordinate so it's not going to be difficult. 

$$ D(\partial_2 L) - \partial_1 L = 0$$
$$ \Rightarrow D(mDs) + mg\frac{1}{\sqrt{h^2d^2\pi^2 + 1}} = 0$$
$$ \Rightarrow mD^2s + mg\frac{1}{\sqrt{h^2d^2\pi^2 + 1}} = 0$$
$$ \Rightarrow D^2s + g\frac{1}{\sqrt{h^2d^2\pi^2 + 1}} = 0$$

This is our answer. Physcially this means the acceleration is proportional to some fraction of gravity. Which makes sense. 

PHASE V

The unspoken step where we actually find the equations of motion. 

$$ D^2 s = -g \frac{1}{\sqrt{h^2d^2\pi^2 + 1}} $$
$$ D s = -g \frac{1}{\sqrt{h^2d^2\pi^2 + 1}} t + v_0 $$
$$ s = -\frac{1}{2} g \frac{1}{\sqrt{h^2d^2\pi^2 + 1}} t^2 + v_0 t + s_0 $$

Which is pretty standard.

*A bead of mass m moves without friction on a triaxial ellipsoidal surface. In rectangular coordinates the surface satisfies $\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2}=1$ for some constands a, b, and c. Identify suitable genrealized coordiantes, formulate a Lagrangian, and find Lagrange's equations.*

PHASE I

The kinetic energy of the system is $\frac12mv^2$

There is no potential energy. This thing is entirely free to roll along the surface. We wonder what kind of crazy spiral geodesics could be made...

PHASE II

An ellipsoid is just a squashed and stretched sphere, so naturally we can reason out that we'll have two coordinates that are angular. Let's call them $\theta$ and $\phi$. Let $\theta$ go from an arbitrary pole to another arbitrary pole, thus having a range from 0 to $\pi$, while $\phi$ goes around perpendicularly, a range of 0 to $2\pi$.

The problem we have now is how do we extract a velocity out of these angles, the ellipsoid is very much an awkard shape. 

Let's think about what, physically, the ellipsoid represents in cartesian to get this figured out. We have a round object with three "semi-major axes," a, b, and c in the x, y, and z directions respectively. Along any of these axes our ellipsoid is symmetric, so let's just declare the z-axis to be the locaiton of our poles since we can do that without loss of genreality. $\theta$ now measures angular distance from the z-poles.

$$ v = \sqrt{(Dx)^2 + (Dy)^2 + (Dz)^2} $$

This is the cartesian velocity. If we could genreally state Dx, Dy, and Dz in the angular coordinates we have selected, we would be good to go. 

The z direction is the easiest, since changing $\phi$ does not change z at all, thus z-related quantities depend only on $\theta$. It's clearly dependent upon a cosine function, as the poles are the edges and the equator is zero.  $cos(\theta)$ is not the whole story though, since z does not range from -1 to 1, but rahter -c to c. This is simply fixed by adding the factor. Thus, 

$$ z = c cos\theta $$

x and y are not so easy, though. They do follow similar relations at the equator if we start $\phi=0$ at positive x...

$$ x = acos\phi, y = bsin\phi $$

This pattern holds at other places besides the equator, but only by a ratio. However, we can note that this ratio is "1" at the equator and "0" at the z-poles, so it relates to a simple sine function again, just based on the other angle.

$$ x = acos\phi sin\theta, y = bsin\phi sin\theta $$

So now we have the coordinates. But not the velocities. Well, we can take time derivatives of all of them...

$$ Dz = -c sin \theta D\theta $$
$$ Dx = acos\phi cos\theta D\theta - a sin\phi D \phi sin\theta $$
$$ Dy = b sin\phi cos\theta D\theta + b cos\phi D\phi sin\theta$$

Which means we can finally write out the velocity.

$$ v = \sqrt{c^2 sin^2\theta (D\theta)^2 + a^2(cos^2\phi cos^2\theta (D\theta)^2 - 2 cos\phi cos\theta sin\phi sin\theta D\theta D\phi + sin^2\phi sin^2\theta (D\phi)^2) \\
+ b^2(sin^2\phi cos^2\theta (D\theta)^2 + 2sin\phi cos\theta cos\phi sin\theta D\theta D\phi + cos^2\phi sin^2\theta (D\phi)^2)} $$

This can, obviously, be simplified.

$$ = \sqrt{c^2 sin^2\theta (D\theta)^2 + a^2(cos^2\phi cos^2\theta (D\theta)^2 + sin^2\phi sin^2\theta (D\phi)^2)+ b^2(sin^2\phi cos^2\theta (D\theta)^2 + cos^2\phi sin^2\theta (D\phi)^2)} $$

At this point it sure *looks* like there should be some $sin^2+cos^2=1$ cancelations, but a is not equal to b and the terms with the same cosntants have different time derivatives tacked onto the end, so in the end all of them are completely unique and do not cancel. 

PHASE III

The Lagrangian is just the kinetic energy as we have no potential energy to speak of.

$$ L = \frac12m(c^2 sin^2\theta (D\theta)^2 + a^2(cos^2\phi cos^2\theta (D\theta)^2 + sin^2\phi sin^2\theta (D\phi)^2)+ b^2(sin^2\phi cos^2\theta (D\theta)^2 + cos^2\phi sin^2\theta (D\phi)^2)) $$

That does get rid of our annoying square root though. 

PHASE IV

Once again, bring in our friend the E-L equation.

$$ D(\partial_2 L) - \partial_1 L = 0$$

Seeing as we have two coordiantes, we will have to do this twice. First, $\theta$.

$$ \Rightarrow D(D\theta m (c^2sin^2\theta + a^2cos^2\phi cos^2\theta + b^2 sin^2\phi cos^2\theta))\\
- m[c^2 sin\theta cos\theta (D\theta)^2 \\
+ a^2( - cos^2\phi cos\theta sin\theta (D\theta)^2 + sin^2\phi cos\theta sin\theta (D\phi)^2)\\
+ b^2(- sin^2\phi cos\theta sin\theta (D\theta)^2 + cos^2\phi sin\theta cos\theta (D\phi)^2)] = 0$$

We still can't simplify, $cos^2x - sin^2x = cos(2x)$, but each term has its own separate derivative and coefficient abc. 

$$ \Rightarrow m [D^2\theta (c^2sin^2\theta + a^2cos^2\phi cos^2\theta + b^2 sin^2\phi cos^2\theta) \\
+ D\theta (2c^2sin\theta cos\theta D\theta \\
- 2a^2cos\phi sin\phi cos^2\theta D\phi- 2a^2cos^2\phi cos\theta sin\theta D\theta \\
+ 2b^2 sin\phi cos\phi cos^2\theta D\phi- 2b^2 sin^2\phi cos\theta sin\theta D\theta) ]\\
- m[c^2 sin\theta cos\theta (D\theta)^2 \\
+ a^2( - cos^2\phi cos\theta sin\theta (D\theta)^2 + sin^2\phi cos\theta sin\theta (D\phi)^2)\\
+ b^2(- sin^2\phi cos\theta sin\theta (D\theta)^2 + cos^2\phi sin\theta cos\theta (D\phi)^2)] = 0$$

At the very least we can pull out the mass and simplify from there.

$$ \Rightarrow D^2\theta (c^2sin^2\theta + a^2cos^2\phi cos^2\theta + b^2 sin^2\phi cos^2\theta) \\
+ D\theta (2c^2sin\theta cos\theta D\theta \\
- 2a^2cos\phi sin\phi cos^2\theta D\phi- 2a^2cos^2\phi cos\theta sin\theta D\theta \\
+ 2b^2 sin\phi cos\phi cos^2\theta D\phi- 2b^2 sin^2\phi cos\theta sin\theta D\theta) \\
-c^2 sin\theta cos\theta (D\theta)^2 \\
- a^2( - cos^2\phi cos\theta sin\theta (D\theta)^2 + sin^2\phi cos\theta sin\theta (D\phi)^2)\\
- b^2(- sin^2\phi cos\theta sin\theta (D\theta)^2 + cos^2\phi sin\theta cos\theta (D\phi)^2) = 0$$

Now, finally, a *couple* of terms cancel. 

$$ \Rightarrow D^2\theta (c^2sin^2\theta + a^2cos^2\phi cos^2\theta + b^2 sin^2\phi cos^2\theta) \\
+ (c^2sin\theta cos\theta (D\theta)^2 \\
- 2a^2cos\phi sin\phi cos^2\theta D\phi D\theta- a^2cos^2\phi cos\theta sin\theta (D\theta)^2 \\
+ 2b^2 sin\phi cos\phi cos^2\theta D\phi D\theta - b^2 sin^2\phi cos\theta sin\theta (D\theta)^2) \\
- a^2(sin^2\phi cos\theta sin\theta (D\phi)^2)\\
- b^2(cos^2\phi sin\theta cos\theta (D\phi)^2) = 0$$

Which we can state in terms of the acceleration.

$$ \Rightarrow D^2\theta \\
 = [- (c^2sin\theta cos\theta (D\theta)^2 \\
+ 2a^2cos\phi sin\phi cos^2\theta D\phi D\theta + a^2cos^2\phi cos\theta sin\theta (D\theta)^2 \\
- 2b^2 sin\phi cos\phi cos^2\theta D\phi D\theta + b^2 sin^2\phi cos\theta sin\theta (D\theta)^2) \\
+ a^2(sin^2\phi cos\theta sin\theta (D\phi)^2)\\
+ b^2(cos^2\phi sin\theta cos\theta (D\phi)^2)] \frac{1}{(c^2sin^2\theta + a^2cos^2\phi cos^2\theta + b^2 sin^2\phi cos^2\theta)}$$

As far as we can tell there's no way to simplify this further as a, b, and c can all be unique. We are notably just asked for Lagrange's Equations here *not* the equations of motion, so we are fortunate. 

So now we do the exact same thing for $\phi$. 

$$ D(D\phi m (a^2 sin^2\phi sin^2\theta + b^2 cos^2\phi sin^2\theta)) \\
- m (- a^2 cos\phi sin\phi cos^2\theta (D\theta)^2 + a^2 sin\phi cos\phi sin^2\theta (D\phi)^2 +b^2 sin\phi cos\phi cos^2\theta (D\theta)^2 - b^2 cos\phi sin\phi sin^2\theta (D\phi)^2)= 0$$



$$ \Rightarrow m(D^2\phi (a^2 sin^2\phi sin^2\theta + b^2 cos^2\phi sin^2\theta)) + m(D\phi (2a^2 sin\phi cos\phi sin^2\theta D\phi - 2b^2 cos\phi sin\phi sin^2\theta D\phi + 2a^2 sin^2\phi sin\theta cos\theta D\theta + 2b^2 cos^2\phi sin\theta cos\theta D\theta)) \\
- m (- a^2 cos\phi sin\phi cos^2\theta (D\theta)^2 + a^2 sin\phi cos\phi sin^2\theta (D\phi)^2 +b^2 sin\phi cos\phi cos^2\theta (D\theta)^2 - b^2 cos\phi sin\phi sin^2\theta (D\phi)^2)= 0$$

$$ \Rightarrow D^2\phi (a^2 sin^2\phi sin^2\theta + b^2 cos^2\phi sin^2\theta) + D\phi (2a^2 sin\phi cos\phi sin^2\theta D\phi - 2b^2 cos\phi sin\phi sin^2\theta D\phi + 2a^2 sin^2\phi sin\theta cos\theta D\theta + 2b^2 cos^2\phi sin\theta cos\theta D\theta) \\
+ a^2 cos\phi sin\phi cos^2\theta (D\theta)^2 - a^2 sin\phi cos\phi sin^2\theta (D\phi)^2 - b^2 sin\phi cos\phi cos^2\theta (D\theta)^2 + b^2 cos\phi sin\phi sin^2\theta (D\phi)^2)= 0$$

$$ \Rightarrow D^2\phi (a^2 sin^2\phi sin^2\theta + b^2 cos^2\phi sin^2\theta) + D\phi (a^2 sin\phi cos\phi sin^2\theta D\phi - b^2 cos\phi sin\phi sin^2\theta D\phi + 2a^2 sin^2\phi sin\theta cos\theta D\theta + 2b^2 cos^2\phi sin\theta cos\theta D\theta) \\
+ a^2 cos\phi sin\phi cos^2\theta (D\theta)^2 - b^2 sin\phi cos\phi cos^2\theta (D\theta)^2= 0$$

$$ \Rightarrow D^2\phi \\
= [- D\phi (a^2 sin\phi cos\phi sin^2\theta D\phi - b^2 cos\phi sin\phi sin^2\theta D\phi + 2a^2 sin^2\phi sin\theta cos\theta D\theta + 2b^2 cos^2\phi sin\theta cos\theta D\theta) \\
- a^2 cos\phi sin\phi cos^2\theta (D\theta)^2 \\
+ b^2 sin\phi cos\phi cos^2\theta (D\theta)^2] \frac{1}{(a^2 sin^2\phi sin^2\theta + b^2 cos^2\phi sin^2\theta)}$$

And once again it doesn't look like we can go any further. 

In short, this is a mess. Were this a sphere a=b=c and thus so many things would *cancel*, but we don't get to have that today, we're on an arbitrary ellipsoid, so of course it's messy.

*The two-bar linkage shown in figure 1.3 (not reproduced) is constrained to move in the plane. It is composed of three small massive bodies interconnected by two massless rigid rods in a uniform gravitaitonal field with vertical acceleration g. The rods are pinned to the central body by a hinge that allows the linkage to fold. The system is arranged so that the hinge is completely free; the members can go through all configurations without collision. Formulate a Lagrangian that describes the system and find the Lagrange equations of motion. Use the computer to do this, because the equations are rather big.*

Now as this is a coding problem we have it provided for us, but we shall still discuss what exactly is going on in detail.

The linkage is subject to gravity but nowhere is it mentioned that any of the masses are actually fixed. Since the equations of motion would be trivial if all three objects are under the influence of gravity equally, we would assume that the problem intends to imply that the central mass (m2) is pinned. Otherwise we can know from physics that an object in a uniform gravitaitonal field will enterfreefall and be indistinguishible from an object not in freefall. 

That said, the program provided for this problem decidedly doesn't do that. One of the equations it produces is clear that all three of the masses exhibit acceleration. We'll get to that later, so for some reason we aren't using the simplificaiton of things in a uniform gravitaitonal field experiencing freefall to be independent of position. Bizarre choice, since it makes it much more complciated, but here we go. 

PHASE I

We can still write down the kinetic and potential energies just fine. 

$$ K = \frac{1}{2} (m_1v_1^2 + m_2v_2^2 + m_3v_3^2) $$

$$ U = mg(h_1+h_2+h_3) $$

PHASE II

Chose a coordinate system. The program picks out x, y, $\phi_1$, and $\phi_3$. Why this set up? 

We really do think x and y are irrelevant, due to us only really caring about coordinates in relation to each other. That said, assuming we do care about the overall position, x and y are the center of mass of the system, which does reduce the three positions of the objects into one. (Though honestly we would rather just fix a coordinate system or the center of mass at a singular mass, say m_2). 

The angles are rather obvious though, they are the angles of mass 1 with respect to mass 2, and mass 3 with respect to mass 2. This is why there is no $\phi_2$ in the list. We would have done this the same way as well.

Now, normally in this step we would try to write out the energies in our coordinate system. We could do that, but it would be... a very, very long process and we are a little low on time, so we'll just outline the procedure. 

The three heights can be found if we let t=0 mean y=0 for m_2, and then m_1 and m_2 will have starting heights based on the angles and the lengths of the linkages. The heights do not change with x, but they will with y. So each h coordinate is found.

The velocities, though, are the actually difficult ones. Each mass has its own velocity. The center of mass velocity is what it all is based on, and so Dx and Dy are the "base" for the velocities. However, the angular accelerations also matter, and the problem is that x and y are not at m_2. Which means we'll have to relate the center of mass to the angles, which we can do, but it would be a long equation. Very long. 

How long? Well, PHASE III should show that pretty clearly...

PHASE III

Here's the Lagrangian printed by the program.

Which probably prints too small to see, and even then it's cut off at the end. But there it is, in all its glory. 

Naturally at this point we would apply the E-L Equations, which, uh...

PHASE IV

These are the the Lagrange Equations for x, y, $\phi_1$ and $\phi_3$. We can see that the equations for x and y are rather to be expected, as the acceleration in x has to drag all three masses along. The y direction is similar, but the g term is added to it as well. Note how the gravity terms are entirely separate from the accelreations based on y, which we do think means we could have completley ignored the effects of gravity, all three masses are influenced the same.

What is harder to understand quickly is what the angles are doing because those equations fill basically the entire board. However, one thing we do note is that *nowhere* in those equations is a g mentioned at all. Once again, supporting our assertion that it is completley irrelevant. The other thing we can determine is that it's very sensitive to the angles given how complicated it is. Which is to be expected, just imagine a three-mass linkage vibrating around, that motion is not only complicated it's clearly related to the known high-chaos motion of a double pendulum.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

pg 53

*Consider a pendulum of length l attached to a support that is free to move horizontally, as shown in figure 1.4 (not reproduced). Let the mass of the support be $m_1$ and the mass of the pendulum bob be $m_2$. Formulage a Lagrangian and derive Lagrange's equations for this system.*

PHASE I:

Kinetic energy: $\frac12 (m_1v_1^2 + m_2v_2^2)$

Potential Energy: $m_2gh$. We can set the potential energy of $m_1$ to zero as it remains at the same height regardless, so this would be the height of $m_2$ alone. 

Are we jumping ahead? Hard to tell, setting a potential zero is part of defining potential enregy, so..... mabye? 

PHASE II

What's a good coordinate system. We think... x and $\theta$, describing the displacement of the support and the angle of the pendulum relative to that support. 

So, the height is easy to determine here, $h = -l cos\theta$. Easy enough.

Now we need to concern ourselves with the velocities. $v_1$ is easy, it's equal to $\dot x$, as this is still a cartesian coordinate. 

$v_2$ will take a little more thought, as it moves with the arc of the pendulum *and* the pivot. We already know the y position of the pendulum bob, it is purely determined by the angle, $- l cos\theta$. The x position is dependent both upon the x coordinate *and* the angle. In this case it is $x + l sin\theta$. 

So with that, we can find velocities. If the y position is $-l cos\theta$ then the time derivative gives us $ l sin\theta \dot \theta$.

Similarly, if hte x position is given by $x + lsin\theta$ then the time derivative is $\dot x + lcos\theta \dot \theta$.

The overall velocity would then be...

$$ v_2^2 = v_x^2 + v_y^2 = \dot x^2 + lcos\theta \dot \theta \dot x + l^2cos^2\theta \dot \theta^2 + l^2sin^2\theta \dot \theta^2 $$

The trig functions add to 1.

$$ v_2^2 = \dot x^2 + lcos\theta \dot \theta \dot x + l^2\dot \theta^2 $$

And so our new kinetic and potential energies are...

$$ \frac12 (m_1\dot x^2 + m_2\dot x^2+ m_2lcos\theta \dot \theta \dot x + m_2l^2\dot \theta^2) $$
$$ -m_2glcos\theta $$

Note that we apparently decided to use dot notation instead of D notation... so long as we're consistent in the problem we should be okay. 

PHASE III

The Lagrangian!

$$ \frac12 (m_1\dot x^2 + m_2\dot x^2+ m_2lcos\theta \dot \theta \dot x + m_2l^2\dot \theta^2) + m_2glcos\theta $$

That was easy. 

PHASE IV

Time to apply our E-L Equation. Since we have two coordinates we have to do it twice. Let's do it with respect to x first.

$$ D(\partial_2 L) - \partial_1 L = 0 $$

$$ D(m_1 \dot x + m_2 \dot x + \frac12 m_2 l cos\theta \dot \theta) - 0 = 0 $$
$$ \Rightarrow m_1 \ddot x + m_2 \ddot x + \frac12 m_2 l cos\theta \ddot \theta - \frac12 m_2 l sin\theta \dot \theta^2 = 0 $$
$$ \Rightarrow (m_1 + m_2) \ddot x = - \frac12 m_2 l cos\theta \ddot \theta + \frac12 m_2 l sin\theta \dot \theta^2 = 0 $$
$$ \Rightarrow \ddot x = \frac12 m_2 l \frac{- cos\theta \ddot \theta + sin\theta \dot \theta^2}{m_1 + m_2} = 0 $$

Once again we are asked for Lagrange's Equations and not the Equations fo MOtion, so we stop here, moving onto the angle.



$$ D(\frac12 m_2 l cos\theta \dot x + m_2 l^2 \dot \theta) + m_2glsin\theta = 0$$
$$ \Rightarrow - \frac12 m_2 l sin\theta \dot x \dot \theta + \frac12 m_2 l cos\theta \ddot x + m_2 l^2 \ddot \theta + m_2glsin\theta = 0 $$
$$ \Rightarrow m_2 l^2 \ddot \theta = - m_2glsin\theta + \frac12 m_2 l sin\theta \dot x \dot \theta - \frac12 m_2 l cos\theta \ddot x $$
$$ \Rightarrow \ddot \theta = - \frac{g}{l}sin\theta + \frac12 \frac{1}{l} sin\theta \dot x \dot \theta - \frac12 \frac{1}{l} cos\theta \ddot x $$

We note that if the x coordinate is not changing we devolve to

$$ \ddot \theta = -\frac{g}{l} sin\theta $$

Which is what we would expect, given that this is apendulum. 

Anyway, we now have Lagrange's Equations. Not the Equations of Motion, but fortunately that's not our concern right now.

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$

PHASE I

What are our energies?

Potential and kinetic, only one actual object. $\frac12 mv^2$ and $mgh$. Nothing special.

PHASE II

One might thing there are two coordinates here, the angle of the bob with its pivot and the angle of the pivot with the center, but this is not the case! We can define this supposed "other angle" as a function of time, since we are given $\omega$, and thus it's rigidly predictable and our only coordinate is... $\phi$

That said finding the values for h and v aren't going to be all that simple. 

We are told that at t=0, the pivot is to the right of the origin. Let's go ahead and declare this point as potential 0. The height will then be the sum of the difference from this point due to the pendulum and the point due to being driven. In other words...

$$ h = -lcos\phi + Rsin(\omega t) $$

Now... the velocity. The velocity will be significantly harder to actually find. However, we can consider it additive again. The velocity of the pivot in x and y is given by $v = - R\omega sin(\omega t) \hat x + R\omega cos(\omega t) \hat y$. These are the *velocities* of the pivot, not the coordinates. They can just be added to the component velocities of the bob.

The bob's relative velocity with respect to the pivot is what we want now, and we can get those from the relative coordinates. 

$$ y = -l cos\phi ; v_y = lsin\phi \dot \phi$$
$$ x = l sin \phi ; v_x = lcos\phi \dot \phi$$

This will make the total velocity be...

$$ v_{total} = (lcos\phi \dot \phi - R\omega sin(\omega t))\hat x + (lsin\phi \dot \phi + R\omega cos(\omega t)) \hat y $$

Now this is in component language. We need to *square* it to get our *actual* value.

$$ v^2 = v_x^2 + v_y^2 = l^2cos^2\phi\dot\phi^2 + R^2\omega^2sin^2(\omega t) - lcos\phi\dot\phi R \omega sin(\omega t) + l^2sin^2\phi^2 \dot\phi^2 + R^2\omega^2cos^2(\omega t) + lsin\phi \dot\phi R \omega cos(\omega t) $$
$$ = l^2\dot\phi^2 + R^2\omega^2 + l\dot\phi R\omega (- cos\phi sin(\omega t) + sin\phi cos(\omega t)) $$

From which we can get the proper kinetic energy.

$$ \frac12 m [l^2\dot\phi^2 + R^2\omega^2 + l\dot\phi R\omega (- cos\phi sin(\omega t) + sin\phi cos(\omega t))]  $$

Our potential energy is

$$ mg(-lcos\phi + Rsin(\omega t)) $$

PHASE III

The Lagrangian becomes

$$ \frac12 m [l^2\dot\phi^2 + R^2\omega^2 + l\dot\phi R\omega (- cos\phi sin(\omega t) + sin\phi cos(\omega t))] - mg(-lcos\phi + Rsin(\omega t))  $$

PHASE IV

The E_L equation is $D(\partial_2 L) - \partial_1 L = 0$. We only have one coordinate, so...

$$ D(ml^2\dot\phi + lR\omega(-cos\phi sin(\omega t) + sin\phi cos(\omega t)) - (l\dot\phi R\omega(sin\phi sin(\omega t) + cos\phi cos(\omega t)) - mglsin\phi ) = 0$$

How nice that there were a handful of constant termst hat fell out. Even though $\omega t$ depends on t, we still take the partial derivatives first! Of course now we have a time dependent full derivative to take...

$$ \Rightarrow (ml^2\ddot\phi + lR\omega(sin\phi \dot \phi sin(\omega t) -cos\phi cos(\omega t) \omega + cos\phi\dot\phi cos(\omega t) - sin\phi sin(\omega t)\omega) - (l\dot\phi R\omega(sin\phi sin(\omega t) + cos\phi cos(\omega t)) - mglsin\phi ) = 0$$

Let's see, looks like some things might cancel. Yes, in fact, we can get rid of four terms! 

$$ \Rightarrow (ml^2\ddot\phi + lR\omega(-cos\phi cos(\omega t) \omega - sin\phi sin(\omega t)\omega) + mglsin\phi  = 0$$
$$ \Rightarrow ml^2\ddot\phi + lR\omega^2(-cos\phi cos(\omega t) - sin\phi sin(\omega t)) + mglsin\phi  = 0$$
$$ \Rightarrow ml^2\ddot\phi = - lR\omega^2(-cos\phi cos(\omega t) + sin\phi sin(\omega t)) - mglsin\phi $$
$$ \Rightarrow \ddot\phi = - \frac{1}{ml}R\omega^2(-cos\phi cos(\omega t) + sin\phi sin(\omega t)) - \frac{g}{l}sin\phi $$

And this is Lagrange's Equation for the system. Naturally it contains the pendulum term.

$$ \Rightarrow \ddot\phi = - \frac{g}{l}sin\phi $$

As the equation is known to not be analytically solvable, we opt to not find the actual Equations of Motion here. Since an analytic form would not exist anyway.
