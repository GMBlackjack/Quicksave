# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

In cartesian, the infinitesimal length ds is given by

$$ \sqrt{dx^2 + dy^2} = \sqrt{1+(dx/dy)^2} dx $$

We can do something similar here for spherical coordinates. First, we note that the arc length formula is, in genreal, $r\theta$, so if we can find the "length" in terms of an angle only, we can just multiply it by R to get the desired result. So, essnetially, we are looking at a space in $\theta$ and $\phi$. 

The full spherical line element is $d\vec l = dr \hat r + r d\theta \hat \theta + r sin\theta d\phi \hat \phi$. (From Zangwill's Electrodynamics). If we note that we are no longer considering the radius, we can have the line element be $d\theta \hat \theta + sin\theta \hat \phi$. 

Plugging this into the infinitesimal length ds, we have

$$ \sqrt{d\theta^2 + sin^2\theta d\phi^2} = \sqrt{1 + sin^2\theta (d\phi/d\theta)^2} d\theta = \sqrt{1 + sin^2\theta \phi'^2} d\theta $$ 

Now we just put this into our integral...

$$ L = R\int_{\theta_1}^{\theta_2} \sqrt{1 + sin^2\theta \phi'^2} d\theta $$

Which is exactly what we sought to show. 

Addendum: to be extra clear let's re-derive where the line element comes from. 

Physically, moving along the radius r is the same as moving in a single direction in xyz, so the element is just dr.

Moving along just the $\theta$ angle is the arc length formula, so $rd\theta$

The last angle is trickly, if for no other reason than because the order of the angles is often ambiguous. If we want the transformations for the two angles to be the same when one angle is not adjusted, that means that $rd\phi$ occurs at that point. We just need to adjust it. If we take $\theta = \pi/2$ to be the "Equator", and 0 to be a "pole" we see that clearly the actual change due to $\phi$ scales with that angle trigonometrically. Specifically, $sin\theta$, so $rsin\theta d\phi$.

Note: it's just as feasible to define the line element with a cosine function and swap the angles. This is the statement of spherical coordinates that provides the form listed at the start of the problem.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

The line element is somewhat trivial for cylindrical, $  d\rho \hat \rho + \rho d\phi \hat \phi + d z \hat z$. 

Let's assume we're not on the cylinder's ends, because such a path would not be smooth. However, since one of our coordinates, z, does not scale with the radius, we can't just take it outside and add it back later. But we are only varying the angle and z, so the radial distance is irrelevant. From this, we construct...

$$ \sqrt{dz^2 + \rho^2 d\phi^2} = \sqrt{1 + \rho^2 \phi'} dz $$

Which we can insert into our original statement to get...

$$ L = \int_{z_1}^{z_2} \sqrt{1 + R \phi'^2} dz $$

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

Our integrand f is $\sqrt{1 + sin^2\theta \phi'^2}$

The E-L equation is

$$ \frac{\partial f}{\partial \phi} - \frac{d}{d\theta} \frac{\partial f}{\partial \phi'} = 0 $$

As mentioned in the problem statmeent, as we are independent of $\phi$ so we reduce this to

$$ - \frac{d}{d\theta} \frac{\partial f}{\partial \phi'} = 0 $$

In order for this to be true, the following must also be true.

$$ \frac{\partial f}{\partial \phi'} = C $$

We have suppressed the sign as it equaled zero before, and C could have any sign. This derivative can actually be taken. 

$$ \frac{sin^2\theta \phi'}{\sqrt{1 + sin^2\theta \phi'^2}} = C $$

Now we just solve for $\phi'$. 

$$ \Rightarrow sin^4\theta \phi'^2 = C^2 + C^2sin^2\theta \phi'^2 $$
$$ \Rightarrow sin^4\theta \phi'^2 - C^2sin^2\theta \phi'^2 = C^2 $$
$$ \Rightarrow \phi'^2 = \frac{C^2}{sin^4\theta - C^2sin^2\theta} $$
$$ \Rightarrow \phi' = \sqrt{\frac{C^2}{sin^4\theta - C^2sin^2\theta}} $$

The next usually clear step is to integrate and then find a function of $\phi$ in terms of $\theta$.

$$ \Rightarrow \phi = \int \sqrt{\frac{C^2}{sin^4\theta - C^2sin^2\theta}} d\theta $$

However, we wish to not do this integral and just justify that C=0. Spherical coordinates can be roatated arbitrarily, so we can always set one point at a pole. Once that is the case, we can still rotate around the polar axis and move the second point to have whatever angle we want. 

If this case, the entire length integral reveals something:

$$ L = R\int_{\theta_1}^{\theta_2} \sqrt{1 + sin^2\theta \phi'^2} d\theta $$

We use our value of $\phi'$ and get...

$$ L = R\int_{\theta_1}^{\theta_2} \sqrt{1 + \frac{C^2}{sin^2\theta - C^2}} d\theta $$

Keep in mind that this is the integral we wish to minimize. Since we can always set it so the $\phi$ angle between the two points is the same, then we are free to set C to whatever we want, as no matter what we do we just need them to be the same at both ends. The minimizing way to do this is clearly to cet C=0. So if C is zero, then we say that $\phi = D$, where D is just some constant. Thus, the shortest path between two points on a sphere is just the arc length!

$$ L = R\int_{\theta_1}^{\theta_2} 1 d\theta = R\Theta$$

Which is exactly what a great circle is, for if we continued the path, we would just circle around the circle. We can adjust $\theta_2$ to be literally anywhere we want.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

These are going to be conic sections, it's how cones work.

We can state the line element in terms of $\phi$ and $\rho$. 

When one moves in $\rho$ what they're actually doing is moving along r, just at a scaled rate. If $\lambda=0$, this rate would be 0, and $\rho = r$. Constructing a triangle, we find $dr = \sqrt{d\phi^2 + \lambda^2d\phi^2}$ or $\sqrt{1+\lambda}d\rho$, which is our line element. 

The other line element segment is simply $\rho d\phi$

So now we've described the line element restricted to a conic. Now we seek to find the integrand. This is...

$$ \sqrt{(1+\lambda) d\rho^2 + \rho^2 d\phi^2} = \sqrt{(1+\lambda)+ \rho^2 \phi'^2 } d\rho $$

The root is our integrand. Now we apply calculus of variations.  

$$ \frac{\partial f}{\partial\phi} - \frac{d}{dr} \frac{\partial f}{\partial \phi'} $$

As usual, we do not depend on $\phi$ and so the $\phi'$ derivative must be constant.

$$ \frac{df}{d\phi'} = C $$

At this point the actual derivative is pretty standard.

$$ \frac{\rho^2\phi'}{\sqrt{(1+\lambda)+ \rho^2 \phi'^2 }} = C $$

We will now re-arrange this...

$$ \Rightarrow \rho^4 \phi'^2 = C^2 + C^2 \lambda + C^2 \rho^2 \phi'^2 $$
$$ \Rightarrow \phi'^2(\rho^4 - C^2 \rho^2) = C^2(1+\lambda) $$
$$ \Rightarrow \phi'^2 = \frac{C^2(1+\lambda)}{\rho^4 - C^2 \rho^2} $$
$$ \Rightarrow \phi' = \sqrt{\frac{C^2(1+\lambda)}{\rho^4 - C^2 \rho^2}} $$


let's see what happens when we just take the integral.

$$ \phi = \sqrt{(1+\lambda)} tan^{-1}\left( \frac{\sqrt{\rho^2-C^2}}{C} \right) + D $$

# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$

First of all, let us consider our coordinate system. We have a spherical situation, but one that's perfectly spherically symmetric, so we can easily consider it just a polar situation. We know we pass through the origin. It is actually completely arbitrary where our second point is, it could be anywhere, and given the symmetry of the problem we can rotate the coordinate system any way we wish and set this point to its radius and any integral. Basically, we seek the shortest path from the center to whatever we want to call the "edge" of our sphere, and then we can simply increase the distance of that edge to infinity. To consider the path before the origin, we just mirror it. 

So, in cartesian coordinates the *time* integral is

$$ \frac{1}{c} \int_{x_1}^{x_2} n \sqrt{1+y'^2} dx $$

We can rather quickly convert this to polar, however since r and $\phi$ are varying this time we can't use the spherical shell version. Fortunately this line element is actually *simpler*, with just dr and r$d\phi$. 

$$ \frac{1}{c} \int_{\phi_1}^{\phi_2} n \sqrt{1+ r^2 \phi'^2} dr $$

Substitute in n...

$$ \frac{1}{c} \int_{\phi_1}^{\phi_2} \frac{a}{r^2} \sqrt{1+ r^2 \phi'^2} dr $$

We can pull the a out, it is irrelevant to the integrand.

$$ \frac{a}{c} \int_{\phi_1}^{\phi_2} \frac{1}{r^2} \sqrt{1+ r^2 \phi'^2} dr $$

With this integrand, our E-L equation is

$$ \frac{\partial f}{\partial \phi} - \frac{d}{dr} \frac{\partial f}{\partial \phi'} = 0$$

WHich, as is normally the case, is not dependent upon $\phi$ so that part goes to zero. Lucky for us this means we can just state the deriative with respect to $\phi'$ is constant and ignore the d/dr derivative, which would have been a little annoying to evaluate. 

$$ \frac{\partial f}{\partial \phi'} = C $$

If we actually take this derivative we get 

$$ C = \frac{r^2 \phi'}{r^2 \sqrt{1+ r^2 \phi'^2}} = \frac{\phi'}{\sqrt{1+ r^2 \phi'^2}}$$

Re-arranging...

$$ C^2(1+r^2\phi'^2) = \phi'^2 $$
$$ \Rightarrow  C^2 = \phi'^2 - C^2r^2\phi^2$$
$$ \Rightarrow  \frac{C^2}{1-C^2r^2} = \phi'^2$$
$$ \Rightarrow  \sqrt{\frac{C^2}{1-C^2r^2}} = \phi'$$

Integrate both sides.

$$ sin^{-1}(Cr) + D= \phi \text{ for } C > 0$$
$$ -sin^{-1}(Cr) + D= \phi \text{ for } C < 0$$
$$ D = \phi \text{ for } C = 0$$

Now if we let D=0 by assigning the angle of the point we're goign for *to* zero, then these simplify and we can put it in terms of r. 

Now if we do this in terms of r...

$$ r = \frac{sin\phi}{|C|} $$

This covers all cases except C=0 which is a discontinuity anyway. 

It is somewaht well known that $r = sin\phi$ is a circle with an edge at the origin in polar coordinates. The C just scales it to a particular size. 

So the leat time-consuming path through something of this index of refraction is a semicircle. This means the least time-consuming path to any point on the edge of the sphere is a path that takes it back to the center, just barely scraping the edge of the sphere! 

Light moves faster in lower indeces of refraction, so near the start of the circle it will be traveling the fastest, and at the edge it will be traveling the slowest. We can think of this sitation like a lot of different shells of indeces of refraction stacked one on top of the other. As we enter higher indices of refraction, the path of least resistance bends at sharper and sharper angles. Now, if we were in a rectangular medium, this would just continue on until it reached an exponential-like curve of some kind. However, since our medium is circular, every time we "turn" at a harsher angle, we also adjust our heading relative to the center. Thus, instead of just approaching an exponential, we can perform an entire arc until we reach the edge of our loop, at which point the light starts moving *back*. The reverse path is still the shortest path, the only difference is that now we are moving from high index to low index, and thus an infinite loop is produced.

# Problem 6 \[Back to [top](#toc)\]
$$\label{P6}$$

The problem seems to have predited that we would try a coordinate transformation. (By proving one dimension we could prove all three). But let's go ahead and do it this way. 

With the three parametric quantities in cartesian we can create the infinitesimal line segment:

$$ ds = \sqrt{dx^2+dy^2+dz^2} = \sqrt{du^2((dx/du)^2+(dy/du)^2+(dz/du)^2)} = \sqrt{x'^2+y'^2+z'^2} du $$

The radical is our function f which we apply to the given EL equaitons. Note that f does not rely on either x, y, or z, so all of the derivatives in relation to x', y', and z' must be constants.

$$ \frac{\partial f}{\partial x'} = C_1 ; \frac{\partial f}{\partial y'} = C_2 ; \frac{\partial f}{\partial z'} = C_3 $$

All three of the actual derivatives take very similar form.s

$$ \frac{x'}{\sqrt{x'^2+y'^2+z'^2}} = C_1 ; \frac{y'}{\sqrt{x'^2+y'^2+z'^2}} = C_2 ; \frac{z'}{\sqrt{x'^2+y'^2+z'^2}} = C_3 $$

Now we take every possible ratio between all of the constants. 

$$ \frac{C_1}{C_2} = \frac{x'}{y'} = \frac{dx}{dy} $$

$$ \frac{C_2}{C_3} = \frac{y'}{z'} = \frac{dy}{dz} $$

$$ \frac{C_3}{C_1} = \frac{z'}{x'} = \frac{dz}{dx} $$

Note that x varies linearly with y, y varies linearly with z, and z varies linearly with x. Thus the shortest path is a straight line in regards to all three variables. (Perhaps it is more accurate to say that we have shown that the projection of the shortest path onto the xy, yz, and zx planes is a straight line, but if we wanted to be pedantic we could argue that since dx/dy is a constant it can be thought of as another axis and *it* could be compared to z, and that compound relation would also be equal to a constant, so any way we slice it, the line is straight).
