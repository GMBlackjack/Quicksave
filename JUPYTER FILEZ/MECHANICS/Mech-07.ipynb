# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

pg 129

*Show that the Moment of Inertia of the following objects are as given:*

*a) The moment of inertia of a sphere of uniform density with mass M and radius R about any line through the center is $\frac25 MR^2$*

Since we're rotating through the center there's no need to go to the matrix, we can just go to the integral.

$$ I = \int r^2 dm $$

The basic version. Naturally we have constant density, but are also integrating over spherical coordinates. Thus, dm is replaced with the density times the volume element. or...

$$ I = \int r^2 \rho r^2 sin\theta drd\theta d\phi $$

But wait! This is actually incorrect, since the r values here are not equivalent! The first one is the radius from the axis, the other is the radius from the center of the sphere. If we want to do this in spherical coordinates we'll need to find an expression for the radius from the axis. This is actually easy enough: $rsin\theta$. So our actual expression is...

$$ I = \int r^2sin^2\theta \rho r^2 sin\theta drd\theta d\phi $$
$$ = \int r^4 sin^3\theta \rho drd\theta d\phi $$

Which we now just integrate as usual over a sphere from radius 0 to R.

$$ = 2\pi \rho \int r^4 sin^3\theta dr d\theta $$
$$ = 2\pi \rho (4/3) \int r^4 dr  $$
$$ = 2\pi \rho (4/3) R^5/5 $$
$$ = (2/5) M R^2 $$

Which is what we sought. 

*b) The moment of inertia of a spherical shell with mass M and radius R about any line through the center $\frac23 MR^2$*

Still going through the center of mass. The actual setup is the same as the previous problem but r is already set to R.

$$ I = \int R^4 sin^3\theta \rho d\theta d\phi $$
$$ = R^4 \rho 2\pi (4/3) $$
$$ = R^2 M (2/3) $$

Which is exactly what we sought. 

Note: technically, we should use $\sigma$ rather than $\rho$ as this is a surface density, but the end result is the same. 

*c) The moment of inertia of a cylinder of uniform density with mass M and radius R about the axis of the cylinder is $\frac12 MR^2$*

Now we get to use cylindrical coordinates! And this time our radius is equal in both pictures, as the distance from the axis is the radius in cylindrical coordinates.

$$ I = \int r^2 \rho r drd\theta dx $$
$$ = 2\pi \int r^3 \rho drdz $$
$$ = 2\pi Z \int r^3 \rho dr $$
$$ = 2\pi Z R^4/4 \rho  $$

Note that Z here is just the height, so...

$$ = \frac12 M R^2 $$

Which is what we sought. 

*d) The moment of inertia of a thin rod of uniform desnity per unit length with mass M and length L about an axis perpendicular to the rod through the center of mass is $\frac{1}{12} ML^2$*

Finally something we can integrate in Cartesian.

$$ I = \int r^2 \rho dxdydz $$

The y and z directions are infinitesimal, so we just pull it out to get

$$ I = \int x^2 \rho dx $$

However, unlike before the integral is now from -L/2 to L/2, so the bounds are not as simple to just imply. Still, the result is...

$$ = \left[ \rho x^3/3 \right]_{-L/2}^{L/2} $$
$$ = \rho(L^3/24 + L^3/24) $$
$$ = \frac{1}{12} ML^2 $$

Which is what we sought.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

*a) the density of a planet increases toward the center. Provide an argument that the moment of inertia of a planet is less than that of a sphere of uniform desnity of the same mass and radius.*

While the total mass is the same, the *position* of the mass matters as well. In fact, moment of Inertia goes like $r^2$ with distance, so a single mass at a further distance has *much* more I the larger it is. 

A sphere that has most of its mass concentrated at the center has much smaller moments of inertia simply because it's at a short distance away from the center. While there is some mass at the far distances, it is much lesser than the uniform density sphere, and the outermost masses provide most of the contribution to the moment of inertia. 

*b) The density as a function of radius inside Jupiter is well approximated by*

$$\rho(r) = \frac{M}{R^3} \frac{sin(\pi r/R)}{4r/R}$$

*where M is the mass and R is the radius of Jupiter. Find the moment of inertia of Jupiter in terms of M and R.*

Let us grab our integral for the moment of inertia of the sphere we used in **Problem 1**.

$$ I = \int r^4 sin^3\theta \rho drd\theta d\phi $$

Of course now $\rho$ is actually a function, so we substitute. 

$$ = \int r^4 sin^3\theta \frac{M}{R^3} \frac{sin(\pi r/R)}{4r/R} drd\theta d\phi $$

$\phi$ and $\theta$ integrate as normal.

$$ = 2\pi (4/3) \int r^4 \frac{M}{R^3} \frac{sin(\pi r/R)}{4r/R} dr $$
$$ = 2\pi (4/3) \frac{M}{R^3} \frac{R}{4} \int r^4 \frac{sin(\pi r/R)}{r} dr $$
$$ = 2\pi (4/3) \frac{M}{R^3} \frac{R}{4} \int r^3 sin(\pi r/R) dr $$

We can take that integral from 0 to R... (yes we used Geogebra)

$$ = 2\pi (4/3) \frac{M}{R^3} \frac{R}{4} \left[ cos(\pi r/R) \frac{-R\pi^3r^3 + 6R^3 \pi r}{\pi^4}+ sin(\pi r/R)\frac{3R^3\pi^2r^2 - 6R^4}{\pi^4} \right]_0^R $$


$$ = 2\pi (4/3) \frac{M}{R^3} \frac{R}{4} \left[ -\frac{-R\pi^3R^3 + 6R^3 \pi R}{\pi^4} \right] $$

$$ = 2\pi (4/3) \frac{M}{R^3} \frac{R}{4} R^4 \pi \left[ \frac{\pi^2 - 6}{\pi^4} \right] $$
$$ = \frac{2}{3} M R^2 \left[ \frac{\pi^2 - 6}{\pi^2} \right] $$

The term in brackets is roughly 0.39. This makes the leading coefficient 0.26, which is in fact less than 2/5, or 0.4.

Which isn't too bad since everything is squared.

$$ I_{xx} = m( 0 + 2/3 + 2/3 + 0 + 1/9 + 1/9 + 1/9 + 1) = m(8/3) $$

So we have...

$$ I_{xy} = I_{yx} = -m (0 - \frac{2}{3\sqrt3} + \frac{2}{3\sqrt3} + 0) = 0 $$

Hold on a sec. This means... that we are already using the principal axes! xyz qualify! *b) A cube of uniform desnity.*

We're actually doing this one first. Easier to think about.

So, first, find the moments of inertia for the Inertia Tensor. 

$$ I_{ii} = \int \zeta_j^2 + \zeta_k^2 dm $$

So in this case we have...

$$ I_{xx} = \int (y^2 + z^2) \rho dzdydz $$

Now let our cube have side length a. Thus, we are going to be integrating all variables from -a/2 to a/2. x first, since it's not present in the actual integrand.

$$ = \rho a \int (y^2+z^2) dydz $$
$$ = \rho a \int [ y^3/3 + yz^2 ]_{-a/2}^{a/2}dz $$

This matches what it should be from a quick look up online. 

$ I_{yy} $ and $ I_{zz}$ are the same by symmetry.

$$ I_{xy} = -\int \rho x y dxdydz $$

Note that we now have functions that are x and y. Which are odd functions. Being integrated from -a/2 to a/2. Everything here goes to zero. For a cube rotating around axes that go through the centers of its faces, there are no products of inertia. Which, well, shouldn't exaclty be surprising given the symmetry. 

This also means that we already have the principal axes, they're just the usual $\hat x, \hat y, \hat z$. Nothing to be all that concerned about.

# Problem 4 \[Back to [top](#toc)\]

*Measure this book. You will admit that it is pretty dense. Don't worry, you will get to throw it later. Show that the principal axes are the lines connecting the centers of opposite faces of the idealized brick approximating the book. Compute the corresponding principal moments of inertia.*

The book weighs 0.86 kg and has dimensions of 25 x 14 x 3.0 cm. These were taken from Amazon rather than actually measured since we did not have a scale and did not want to walk all the way over to the lab. 

Regardless, we don't actually need any of the dimensions to prove that the principal axes are through the faces, this is true for *any* rectangular prism of constant density. (The book is almost entirely paper). For the Products of Inertia are always:

$$I_{ij} = -\int \zeta_i \zeta_j dm$$

But the actual bounds of the integral are what are important, they go from -a/2 to a/2, where a is whatever side we are integrating over. The only forms of variables we get in this function are xy, yz, and xz, all of which are *odd* functions, and integrate to *zero* when going from a negative to positive bound of same magnitude. Thus, all off-diagonal terms, the products of inertia, are *always* zero for any rectangular prism of constant density.

That said we need to find the actual moments of inertia for these principal axes, so let's go ahead and do that. It's similar to what was done for the cube in **Problem 3**, except with numbers.

$$ I_{ii} = \int \zeta_j^2 + \zeta_k^2 dm $$

You know what it'll acutally be easier to do this in variable form and input numbers at the very end. Let the dimensions themselves be X Y Z. The mass is then $M = \rho XYZ $. We can actually even do all three at once if we let ijk be arbitrary... (Yes we really don't want to write this down three times). 

$$ I_{ii} = \int_{-K/2}^{K/2}\int_{-J/2}^{J/2}\int_{-I/2}^{I/2} (j^2+k^2) \rho didjdk $$
$$ = I \rho \int_{-K/2}^{K/2}\int_{-J/2}^{J/2} (j^2+k^2) djdk $$
$$ = I \rho \int_{-K/2}^{K/2} \left[ (j^3/3+jk^2) \right]_{-J/2}^{J/2} dk $$
$$ = I \rho \int_{-K/2}^{K/2} (J^3/12+Jk^2) dk $$
$$ = I \rho (KJ^3/12+JK^3/12) $$
$$ = IJK \rho (J^2/12+K^2/12) $$
$$ = \frac{M}{12} (J^2+K^2) $$ 

Which means our individual moments of inertia are

$$ I_{xx} = \frac{M}{12} (Y^2+Z^2) $$ 
$$ I_{yy} = \frac{M}{12} (X^2+Z^2) $$ 
$$ I_{zz} = \frac{M}{12} (X^2+Y^2) $$ 

We can plug in the measured values now. 

$$ I_{xx} \approx 15 kgcm^2 $$ 
$$ I_{yy} \approx 45 kgcm^2 $$ 
$$ I_{zz} \approx 59 kgcm^2 $$ 

Which means rotating it around its shortest dimension is the hardest. Which makes sense, the mass is distributed the furthest away from that axis in all directions.
