$$Df(t) = n(1-e^2)^{1/2} \left( \frac{a}{R(t)} \right)^2$$

*with*

*Notice that n disappears from the equations if they are written in terms fo a new independent variable $\tau=nt$. Also notice that a and R(t) appear only in the comibnation a/R(t)*

Since we need to actually do some work in this problem, we believe we should actually go through the effort to show that the above formula is true, rather than just accept it. Though depending on what we find in our astrophysics test, this might only be a few lines. 

For an ellipse, the radius R is given by:

$$ R = \frac{a(1-e^2)}{1+ecos\theta} $$

Which matches our second condition, if we replace $\theta$ with f(t). So with just the basic definition of an ellipse, we have shown the a/R(t) relation to be true. 

But now we need to find the derivative with respect to time of this angle f. Kepler's Second law dictates that the rate at which area is swept out is cvonstant, from which we can find this relation:

$$ DA = \frac12 R^2 Df $$

Which we can solve for Df

$$ Df = 2 (DA) \frac{1}{R^2} $$

The problem is we need to find DA now. The usual relation is $DA = L/2\mu$, but will that really help, seeing as the final formula here has neither the reduced mass nor the angular momentum? Well, elsewhere we find $L = \mu \sqrt{GMa(1-e^2)}$. (In the text we drew this from, M is the sum of both masses, which is generally dominated by the primary mass anyway). 

$$ Df = \sqrt{GMa(1-e^2)} \frac{1}{R^2} $$
$$ Df = \sqrt{GMa}\sqrt{(1-e^2)} \frac{1}{R^2} $$

Digging around *online* we found a the exact expression for the mean motion we needed: $ n^2 = \frac{GM}{a^3} $. This comes from Kepler's Third Law.

$$ Df = na^2\sqrt{(1-e^2)} \frac{1}{R^2} $$
$$ \Rightarrow Df = n\sqrt{(1-e^2)} \frac{a^2}{R^2} $$

Which is exactly what we sought.
