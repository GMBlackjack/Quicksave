# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

*A mass $m_1$ rests on a frictionless horizontal table. Attached to it is a string which runs horizontally to the edge of the table, where it passes over a fricitonless, small pulley and down to where it supports a mass $m_2$. Use as coordinates x and y the distances of $m_1$ and $m_2$ from the pulley. These satisfy the constraint equation f(x,y) = x+y=const. Write down the two modified Lagrange equations and solve them (together with the constraint equation for $\ddot x, \ddot y$, and the Lagrange multiplier $\lambda$). Use $\lambda \frac{\partial f}{\partial x} = F_x^{constraint}$ and the correspponding equation for y to determine the tension in the string.*

Just because we have a constraint does not mean we are permitted to avoid the STEPS!

PHASE I:

Energies. Kinetic is $\frac12(m_1v_1^2 + m_2v_2^2)$

Potential is $m_2gh_2$. This is because the horizontal table's potential energy does not change. (Arguably this would require us to declare a constant mgh but that'll just needlessly complciate things in our judgment. We do this without defining coordinates too!)

PHASE II:

Now we come up with our actual coordinates. They are provided, so that solves some of our quesitons. We have x and y, and so $v^2_1 = \dot x^2$ and $v_2^2 = \dot y^2$. It is very convenient that our system here is effectively on rails, each one only moves in a single coordinate. We might be tempted to say $h_2 = y$ as well, but e actually need to be careful, the constraint x+y=c assumes a positive y, so in actuallity $h_2 = -y$. Strictly speaking the velocities will have sign issues as well, but we only have their squared values so we avoid that. 

This is also the part where we identify our equation of constraint. Fortunately for us we've already been given one: x+y=c, where c is some constant. Since the constraint needs to take the form f=0, we just adjust this to x+y-c=0, and we're good.

PHASE III:

The Lagrangian! 

Without the constraint equation, it is...

$$ \frac12(m_1\dot x^2 + m_2 \dot y^2) + m_2gy $$

And when we **add** the equation of constraint...

$$ \frac12(m_1\dot x^2 + m_2 \dot y^2) + m_2gy + \lambda(x + y - c)$$

PHASE IV

Now we get our E-L equation. 

$$  D(\partial_2 L) - \partial_1 L = 0 $$

Which is done for all three coordinates. Three? Yes! $\lambda$ qualifies as one, though when we take it in this case we just end up with the constraint equation back.

$$ m_1\ddot x - \lambda = 0$$
$$ m_2\ddot y - m_2 g - \lambda = 0$$
$$ -x-y+c=0 $$

These are the Lagrange Equations. 

PHASE V

The secret step: acutally find things out with these equations. We seek the value of $\lambda$ so we can find the force in the x direction according to the provided force. 

The constraint can have its derivative taken to reveal that $\dot x = -\dot y$ and $\ddot x = -\ddot y$ which... of course it does, the string isn't elastic. This means we can solve for the acceleration in x and insert it into y's expression.
$$ \ddot x = \frac{\lambda}{m_1} $$
$$ \Rightarrow -\frac{m_2\lambda}{m_1} - m_2g - \lambda = 0 $$
$$ \Rightarrow -\frac{m_2}{m_1} - \frac{m_2g}{\lambda} - 1 = 0 $$
$$ \Rightarrow -\frac{m_2}{m_1} - 1 = \frac{m_2g}{\lambda} $$
$$ \Rightarrow (-\frac{1}{m_1} - \frac{1}{m_2})\frac{1}{g} = \frac{1}{\lambda} $$
$$ \Rightarrow \frac{-g}{\frac{1}{m_1} + \frac{1}{m_2}} = \lambda $$

And so F = $\lambda \frac{\partial f}{\partial x}$. Since the constraint equation f is just y+x-c, this becomes 1.

$$ F_x^{constraint} = \frac{-g}{\frac{1}{m_1} + \frac{1}{m_2}} $$

Same for y, as $\lambda$ is the same and the derivative of y+x-c with respect to y is still 1. Naturally the tension in the string is equal in magnitude for both components. 

$$ F_y^{constraint} = \frac{-g}{\frac{1}{m_1} + \frac{1}{m_2}} $$

Which is to say the force in the tension is some ratio of gravity related to the difference between the masses. Makes sense, as gravity is the primary driver of the forces in the system.

One has to ask--are the forces in the string right? The magnitude makes sense, but what of direction? The force of tension is fighting against gravity and pulling the box in the x position which, in the way we've defined our coordinates, would mean the force is negative--fighting to "shrink" the distance value. Our tension here is in fact negative.

It's also negatie in the y direction even though y is increasing, but that's because gravity is also acting on that point.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

*Consider a mass m that hangs from a string, the other end of which is wound several times around a wheel (radius R, moment of inertia I) mounted on a frictionless horizontal axle. Use as coordiantes for the mass and the wheel x, the distance fallen by the mass, and $\phi$, the angle through which the wheel has turned (both measured from some convenient reference position). Write down the modified Lagrange equations for these two variables and solve them (together with the constraint equation) for $\ddot x$ and $\ddot \phi$ and the Lagrange multiplier. What is the tension force in the string on the mass? Calculate also the quantity $\lambda \frac{\partial f}{\partial \phi} = F_\phi^{constraint}$ -- what does this represent?*

PHASE I

Energies.

We have two kinetic energies, a translational and a rotational.

$$ \frac12 m v^2 + \frac12 I\omega^2 $$

In this case we do not care waht the moment of inertia is made of, as it is provided. As the wheel does not move up or down, we have only a single potential energy: that of the weight.

$$ mgh $$

And these are our energies. 

PHASE II

The coordinate system is in fact provided for us. We are told to use x and $\phi$. Of special note x is the distance fallen, so the h = -x, giving us a sign to keep track of. 

The velocity can be simply written as $v^2 = \dot x^2$.

The angular velocity is just $\dot \phi = \omega$. Simple. 

However, the angular velocity is related to the tangential velocity by $r\omega = v$ so we have $\dot \phi = \dot x / R$. This we can use as our equation of constraint, and it's velocity related so we have to use the velocity version. $\dot \phi - \dot x/R = 0$

We note that there is some ambiguity as to the direction of positive $\phi$. Since velocity is going to be negative, $\omega$ is also going to be megative, so that means in this case "forward" is the negative direction for the wheel. Can't say if this is left or right side since we don't have that orientation provided. 

PHASE III

The Lagrangian is

$$ \frac12m\dot x^2 + \frac12I\dot \phi^2 + mgx $$

And the *modified* Lagrangian for velocity constraints is...

$$ \frac12m\dot x^2 + \frac12I\dot \phi^2 + mgx + \lambda' (\dot \phi - \dot x / R)$$

PHASE IV

Apply the E-L equations for x, $\phi$, and $\lambda'$.

$$D(\partial_2 L) - \partial_1 L = 0$$

$$ m\ddot x- \lambda' / R - mg = 0$$
$$ I\ddot \phi + \lambda' = 0 $$
$$ \dot \phi - \dot x / R = 0 $$

PHASE V

Bonus step! Which is to say we figure out what the problem actually wants. It wants the forces. 

We must note that the actual force equation is not what's given above, it is:

$$ \lambda' \frac{\partial Df}{\partial q^i} = \lambda' \frac{\partial \psi}{\partial q^i} = \lambda' \frac{\partial (\dot \phi - \dot x /R)}{\partial q^i} = F_i $$

Anyway, to find this we need to figure out what $\lambda'$ is, which is going to be some linear funciton of time. We can do this by taking the time derivative of the constraint.

$$ \ddot \phi = \ddot x / R $$

Which we can use to solve for $\ddot x$ twice.

$$ m\ddot x - \lambda ' / R - mg = 0 \Rightarrow \ddot x = \frac{\lambda'}{mR} + g $$
$$ \ddot x = -R \lambda'/I  $$

And so...

$$  \frac{\lambda'}{mR} + g = -R \lambda'/I  $$
$$ \Rightarrow \frac{1}{mR} + \frac{g}{\lambda'} = -R /I  $$
$$ \Rightarrow \frac{1}{\lambda'} = -\frac{R}{gI} -  \frac{1}{gmR} $$
$$ \Rightarrow \lambda' = \frac{g}{-\frac{R}{I} -  \frac{1}{mR}} $$

Thus our forces are:

$$ F_x = \frac{g}{\frac{R}{I} + \frac{1}{mR}} \frac{1}{R} $$

$$ F_\phi = -\frac{g}{\frac{R}{I} + \frac{1}{mR}}  $$

The x direction feels a positive (upward) force that scales with gravity, while the $\phi$ variable is dragged in its negative direction. This is sensible for what we expect tension to be in a string. Specifically, the $F_\phi$ "force" is actually a torque. We note that the moment of inertia has units of mass length length, which effectively turns the denominator for $F_\phi$ into a unit of 1/(mass length), which means we have acceleration mass length, which is force length. Which is exactly what a torque is.

Note the 1/R that removes the extra length unit in $F_x$, making it a standard force.

$$ h_2 = -l_1 cos\theta_1 - l_2 cos\theta_2 $$

The "heights" of both pendulums need to be added. The velocity can probably be constructed in a similar way, but it would be even more of a mess, so instead we are going to define x and y of the second mass in terms of the angles. We already did it for y. But what is x? Well...

$$ x_2 = l_1 sin\theta_1 + l_2 sin\theta_2 $$

$$ v_{x2} = l_1 cos\theta_1 \dot \theta_1 + l_2 cos\theta_2 \dot \theta_2 $$
$$ v_{y2} = l_1 sin\theta_1 \dot \theta_1 + l_2 sin\theta_2 \dot \theta_2 $$

$$v_2^2 = l_1^2cos^2\theta_1 \dot \theta_1^2 + l_2^2 cos^2\theta_2\dot \theta_2^2 + 2l_1l_2cos\theta_1 cos\theta_2\dot \theta_1 \dot \theta_2 + l_1^2 sin^2\theta_1 \dot \theta_1^2 + l_2^2 sin^2\theta_2\dot \theta_2^2 + 2l_1l_2sin\theta_1 sin\theta_2\dot \theta_1 \dot \theta_2$$
