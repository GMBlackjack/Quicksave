# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

*For each of the mechanical systems described below, give the number of degrees of freedom of the configuration space.*

*a) Three juggling pins.*

Each pin has a 3 degrees of freedom in the spatial sense, and 2 in the rotational sense as they are symmetric azimuthally. Each of them are independent of each other, so there's **15** degrees of freedom total. 

*b) A spherical pendulum, consisting of a point mass (the pendulum bob) hanging from a rigid masselss rod attached to a fixed support point. The pendulum bob may move in any direction subject to the constraint imposed by the rigid rod. THe point mass is subject to the unform force of gravity.*

Two degrees of freedom, angles with respect to the anchor of the pendulum. These two angles describe its position perfectly. As the bob is a point mass, it has no rotation degrees of freedom. So a total of **2** degrees of freedom. 

*c) A spherical double pendulum, consisting of one point mass hanging from a rigid massless rod attached to a second point mass hanging from a second massless rod attached to a fixed support point. The point masses are subject to the uniform force of gravity.*

The problem seems a bit ambiguous, are the two pendulums attached to each other or entirely seprate? In both cases, though, the answer is the same. Each pendulum offers 2 degrees of freedom. If they are unattached, this is **4**. If they are attached, while they are related to each other, you can technically have either of the pendulums have their own angles, which would still be **4**. 

*d) A point mass sliding without friction on a rigid curved wire.*

You would have 1 degree of freedom, since you can parameterize the wire's length as a coordiante. **1** degree of freedom. 

*e) A top consisting of a rigid axisymmetric body with one point on the symmetry axis of the body attached to a fixed support, subject to a uniform gravitational force.*

So if the top is fixed to a support, that means it doesn't actually physically move, but we can assume it's allowed to rotate around this point. We can describe the orientation of such a top with two angles: how far from the axis it is, and which direction it is pointing (as a clock's hands would point.) Thus, **2** degrees of freedom. 

*f) The same as e), but not axisymmetric*

Same as e), except we now care how far the top has turned around its spin axis, so three angles are required. (imagine the top above, except it has a green dot on one side. If we want to know where that green dot is, we need another angle.) **3** degrees of freedom.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

*For a free particle an appropriate Lagrangian is*

$$ L(t,x,v) = \frac12 mv^2 $$

*Suppose that x is the constant-velocity straight-line path of a free particle, such that $x_a = x(t_a)$ and $x_b = x(t_b)$. Show that the action on the solution path is*

$$ \frac{m}{2} \frac{(x_b-x_a)^2}{t_b-t_a}$$

The action is the integral over time of the Lagrangian. Since we are on a straight line we don't have to worry at all about the path.

$$ \int_{t_a}^{t_b} \frac{1}{2} mv^2 dt = \frac{m}{2} \int_{t_a}^{t_b} v^2 dt$$

Now, the interesting part here is that the velocity is constant, so it is not changing in time at all. Thus, we end up with:

$$ = \frac{m}{2} v^2 (t_b-t_a) $$

Now velocity itself is just distance over time, which can be represented as x/t, which we can use to substitute and get...

$$ = \frac{m}{2} \left(\frac{x_b-x_a}{t_b-t_a}\right)^2 (t_b-t_a) $$

$$ = \frac{m}{2} \frac{(x_b-x_a)^2}{t_b-t_a} $$

Which is exactly what we sought.

*Derive the Lagrange equations for the following systems, showing all of the intermediate steps as in the harmonic oscillator and orbital motion examples.*

*a) An ideal planar pendulum consists of a bob of mass m connected to a pivot by a massless rod of length l subject to uniform gravitational acceleration g. A Lagrangian is $L(t,\theta,\dot\theta) = \frac12 ml^2 \dot\theta^2 + mglcos\theta$. The formal paramaters of L are t, $\theta$, and $\dot\theta$; $\theta$ measures the angle of the pendulum rod to a plumb line and $\dot\theta$ is the angular velocity of the rod.*

Well we're already given the Lagrangian, so this isn't going to be difficult at all. Only have one variable to worry about to boot. We should get used to our notation, so let's write out the E-L equation:

$$ \partial_1 L - D(\partial_2 L) = 0$$

Where D is implied to be the derivative with respect to time. All we have to do is take the derivatives and we arrive at the Lagrange equations. 

$$ \Rightarrow -mglsin\theta - D(ml^2\dot\theta) = 0 $$
$$ \Rightarrow -mglsin\theta - ml^2\ddot\theta = 0$$

We can get rid of the signs since it all equals zero.

$$ \Rightarrow mglsin\theta + ml^2\ddot\theta = 0$$

And this is our Lagrange equation. 

*b) A particle of mass m moves in a two-dimensional potential $V(x,y) = (x^2+y^2)/2 + x^2y - y^3/3,$ where x and y are rectangular coordinates of hte particle. A Lagrangian is $L(t;x,y; v_x,v_y) = \frac12 m (v_x^2 + v_y^2) - V(x,y)$*

Writing the lagrangian out in full seems like a good idea. 

$$ L =  \frac12 m (v_x^2 + v_y^2) - (x^2+y^2)/2 - x^2y + y^3/3 $$

Once again, E-L equation.

$$ \partial_1 L - D(\partial_2 L) = 0$$

Though this time we need to do it for both x and y. 

$$ -x - 2xy - D(mv_x) = 0 $$
$$ \Rightarrow x + 2xy + ma_x = 0 $$

Where $a_x$ is accelreation in the x direction, or $\ddot y$.

Now for y...

$$ -y - x^2 + y^2 - D(mv_y) = 0 $$
$$ \Rightarrow -y - x^2 + y^2 - ma_y = 0 $$

And that is both of the Lagrange equations. 

*c) A Langrangian for a particle of mass m constrained to move on a sphere of radius R is $L(t;\theta,\varphi;\alpha,\beta) = \frac12 mR^2(\alpha^2 + (\beta sin\theta)^2).$ The angle $\theta$ is the colatitude of the particle and $\varphi$ is the longitude; the rate of change of the colatitude is $\alpha$ and the rate of change of the longitude is $\beta$.*

The Lagrangian is provided again, even though it looks a little odd we'll jsut take it. So we apply the E-L equation agian, once for each angle...

First, $\theta$

$$ mR^2\beta^2sin\theta cos\theta - D(mR^2\alpha) = 0$$

Since we don't have a normal symbol for angular acceleration to use here, let's write $\alpha$ as the time derivative of $\theta$.

$$ \Rightarrow mR^2\beta^2sin\theta cos\theta - mR^2D^2\theta = 0 $$

And as for$\varphi$...

$$ 0 - D(mR^2sin^2\theta \beta) = 0 $$

Now hold on, $\theta$ depends on time! This just became slightly more complicated. The derivative here will need to rely on the product rule, and take a few more steps...

$$ \Rightarrow -mR^2(sin^2\theta D^2\varphi + 2sin\theta cos\theta D\theta D\varphi) = 0 $$
