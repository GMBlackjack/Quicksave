# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$



So the general method for finding eigenvalues and eigenvectors is to subtract the eigenvalues from the diagonal, take the determinant, set it to zero, and find the eigenvalues. After the values have been found, take one at a time and solve for the eigenvectors.

However, we have a unique opportunity here. We know what this unitary matrix IS. It's the rotation operator in 2D space. This allows us to *deduce* the eigenvalues and eigenvectors through reasoning alone. It might seem obvious that the only options for eivenvalues are 1 and -1, but that would be for a mroe specific kind of rotaiton, specifically one TO a certain angle. What we have here is a matrix that takes a vector and rotates it in a *specific* angle, one that can be freely specified. This makes it seem like the eigenvalues change based on the angle and might not exist for some angles at all... which means we do actually have to physically calculate it out. 

Subtract $\lambda$ from the diagonals and take the determinant.

$$ detU = (cos\theta - \lambda)^2 + sin^2\theta = cos^2\theta - 2\lambda cos\theta + \lambda^2 + sin^2\theta = 1 - 2\lambda cos\theta + \lambda^2 $$

Now we set this to zero and solve for the roots. 

$$ 1-2\lambda cos\theta + \lambda^2 = 0 $$

$$ \Rightarrow \lambda = -cos\theta \pm \sqrt{cos^2\theta -1} = -cos\theta \pm \sqrt{-sin^2\theta} $$

This does actually confirm what we suspected, real roots only exist when $sin^2\theta=0$, which just so happens to coincide with the cosine taking values of +1 and -1, the 180 and 0 degree turns. (Which effectively have all possible eigenvectors). 

But the *imaginary* eigenvalues raise some interesting questions. Are there imaginary eigenvectors? Well, if we pull out the imaginary number, we get...

$$ \lambda = -cos\theta \pm isin\theta $$

Which we can represent with euler's formula as $\lambda = -e^{\pm i\theta}$

Eigenvalues, being constant, do not exactly have "magnitudes" but if one considers their real and imaginary parts as components, then yes they can be squared and then added to get 1. Which makes sense since we are basically rotating things around a circle of radius 1. 

Now, the question is, what does this mean to the eigenvectors? Well, since we're dealing with trig functions, we can set up the relations with the trig funciton versions of the eigenvalues:

$$
\begin{bmatrix}
cos\theta & -sin\theta \\
sin\theta & cos\theta \\
\end{bmatrix} \begin{bmatrix}
x \\
y \\
\end{bmatrix} = (-cos\theta \pm isin\theta) \begin{bmatrix}
x \\
y \\
\end{bmatrix}$$



Which gives us the equations...

$$x cos\theta  - y sin\theta = -x cos\theta \pm x i sin\theta $$
$$x sin\theta  + y cos\theta = -y cos\theta \pm y i sin\theta $$

The cosines quickly combine.

$$2x cos\theta  - y sin\theta = \pm x i sin\theta $$
$$x sin\theta  + 2y cos\theta = \pm y i sin\theta $$


We can solve for x

$$x sin\theta  = - 2y cos\theta \pm y i sin\theta $$
$$x = \frac{- 2y cos\theta \pm y i sin\theta}{sin\theta} $$

And then substitute 

$$2(\frac{- 2y cos\theta \pm y i sin\theta}{sin\theta}) cos\theta  - y sin\theta = \pm (\frac{- 2y cos\theta \pm y i sin\theta}{sin\theta}) i sin\theta $$

And we are stubbornly going to solve this in complete generality rather than trying to deal with cases. 

$$2y(\frac{- 2 cos\theta \pm i sin\theta}{sin\theta}) cos\theta  - y sin\theta = \pm y (\frac{- 2 cos\theta \pm i sin\theta}{sin\theta}) i sin\theta $$

Which y can just divide out of! But wait, what does this MEAN? 

It means that while x and y are inexorably related, they can be set to any *size* and still be unafected overall. That is to say, we are free to specify a multiplicative constant on x and y that doesn't change the eigenvector nature. So the relation

$$x = y\frac{- 2 cos\theta \pm  i sin\theta}{sin\theta} $$

Is sufficient for our purposes. A magnitude of the eigenvector is not specified, so while generally we want eigenvectors of magnitude 1, it's not required here, so we can just set y to 1 and get the eigenvectors:

$$
\begin{bmatrix}
\frac{- 2 cos\theta \pm  i sin\theta}{sin\theta} \\
1 \\
\end{bmatrix}
$$

We note that this becomes quite undefined at $\theta=0$. However, that actually makes sense, because no then *every* vector given to the rotation matrix is an eigenvector becuase it's not rotating anything! A similar thing happens at 180 degrees, where *every* vector gets perfectly flipped. 

This can alternatively be written as...

$$
\begin{bmatrix}
\frac{- cos\theta - e^{\pm i\theta}}{sin\theta} \\
1 \\
\end{bmatrix}
$$

Where we don't watch the plus-or-minus too closely since the orientation of it does not matter, there is no order to "which eigenvalue comes first."

The major difference here is that we now have a criteria for the eigenvectors we need: they need to be orthonormal. We do note that basis is still not specified, so they can be whatever length we want.

Anyway, begin the 3x3 determinant taking.

$$ [(2-\lambda)^3 -1 -1] - [ +(2-\lambda) + (2-\lambda) + (2-\lambda) ] = (2-\lambda)^3 - 2 - 3(2-\lambda) $$

$$ = 8 - 12\lambda + 6\lambda^2 - \lambda^3 -8 + 3\lambda = -9\lambda + 6\lambda^2 - \lambda^3 $$

Which we factor into...

$$ = (\lambda)(-9 + 6\lambda -\lambda^2) = (\lambda)(\lambda-3)^2 $$

And yes, we have a duplicate eigenvalue, two 3s, which means that's where our difficulty finding orthonormality will rest. (And we have 0.)

Might as well try to solve the "0" eigenvector first for something that's not a trivial solution. 

$$ 2x + iy + z = 0 $$
$$ -ix + 2y + iz = 0 $$
$$ x - iy + 2z = 0 $$

Let's start combining things to get more relations...

$$ z = -2x-iy $$
$$ -3x - 3iy = 0 $$
$$ x = -iy $$
$$ z = iy $$
$$ -y + 2y + -y = 0 $$

Since that last one is automatically true no matter what y is we are free to specify. Let y=1, our vector becomes (-i,1,i).

So now, the more complicated case...

$$ 2x + iy + z = 3x $$
$$ -ix + 2y + iz = 3y $$
$$ x - iy + 2z = 3z $$

Our bet here is that there are going to be a lot more options here. Similar strategy as before, though...

$$ z = x-iy $$

Which, no matter what one of the other equations we stick it into, creates an "always true" situation. Which means we have full freedom of specificity... so to make them orthogonal, we make a very simple choice: let one option be x=0, and another be y=0. That way they cannot add to each other and must be orthogonal, we'll just have to check to make sure the first one remains orthogonal afterward. 

Anyway, with x=0...

$$ z = -iy $$

Which immediately leads back to a loop. So we let z=1 in order to guarantee differentiation from the original eigenvector, leaving y as i, and the vector (0,i,1) 

With y=0...

$$ z = x $$

Which again leads to a loop. We specify z=x=1, giving us (1,0,1). 

Now, the two eigenvalue-3 eigenvectors have been constructed to be orthogonal, and the eigenvalue-0 should automatically be so. But we should check. Can we combine our eigenvalue-3 eigenvectors to get the (-i,1,i) vector, allowing for imaginary coefficients?

Well, let's try. In order to get -i in the x slot, the coefficient for (1,0,1) has to be -i. In order to get 1 in the second slot, the coefficient there *also* has to be -i. This makes the third slot -i-i or -2i, making the attempt at remaking (-i,1,i) simply (-i,1,-2i) which is not right, so it cannot be made. Check complete, orthogonality conformed.

Vectors are (-i,1,i), (0,i,1), and (1,0,1).
