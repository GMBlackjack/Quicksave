# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Generally we use the CR relations for this, but note that the CR relations are *necessary* not *sufficient* for differentiability (which implies analytic behavior). 

The CR relations are

$$ \frac{\partial u}{\partial x} = \frac{\partial v}{\partial y} $$
$$ \frac{\partial v}{\partial x} = -\frac{\partial u}{\partial y} $$

Where u is the real part and v is the imaginary part. For our equation, we have:

$$ 2x = -2x $$
$$ -2y = -2y $$

This is NOT correct, the sign flips are on the wrong side, so the function is NOT analytic everywhere, so we do not need to differentiate the function. 

Notably the flipping of the sign adjusts things properly.

$$ 2x = 2x $$
$$ -2y = 2y $$

Which is proper. Keep in mind, though, this doesn't necessarily mean that we *are* analytic, just that we meet the initial requirement. The *sufficient* condition is that all the partial derivatives are continuous, which everything clearly is. So we have shown the analytic nature. Now we must actually calculate the derivative.

Anyway, we note that for the derivative z = x+iy. Can we rewrite the funciton in terms of z? Why, yes! it's pretty clearly a polynomail setup:

$$ x^2 + 2ixy - y^2 = (x+iy)^2 = z^2 $$

And we can just differentiate that directly. 

$$ f'(z) = 2z = 2x + 2iy$$

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

NOTE: the minus should be a plus, as communicated over email.

$$ u(x,y) = ln(x^2+y^2) $$

Harmonic equaitons are ones that satisfy Laplace's equation, that is, the square gradient is zero. $\nabla^2 u = 0$. 

If we take derivatives with respect to x first, we get this sequence:

$$ ln(x^2+y^2) \rightarrow \frac{2x}{x^2+y^2} \rightarrow \frac{2}{x^2+y^2} + \frac{-4x^2}{(x^2+y^2)^2} = \frac{-2x^2 + 2y^2}{(x^2+y^2)^2} $$

Symmetry will produce the exact same thing in y, except with the negative sign on the y term instead.

$$ ln(x^2+y^2) \rightarrow \frac{2y}{x^2+y^2} \rightarrow \frac{2x^2 - 2y^2}{(x^2+y^2)^2} $$

If we add the two second derivatives togehter, we get zero. Therefore, the function is harmonic. 

Basically, we assume the u given to us and the v we seek follow the CR relations. Which means:

$$ \frac{\partial v}{\partial y} = \frac{2x}{x^2+y^2} $$
$$ \frac{\partial v}{\partial x} = \frac{-2y}{x^2+y^2} $$

Unfortunately taking an integral is not so easy as taking a derivative, so where to go form here is not immediately obvious. 

Okay so the key insight (which we never would have found without computer help) is that this is close to the form of a known funciton, $\frac{d}{dx} tan^{-1}(x) = \frac{1}{1+x^2} $. We can re-arrange our integral into something close to this form:

$$ v = \frac{2}{x}\int \frac{1}{1+y^2/x^2} dy + f(x) = \frac{2}{x}x tan^{-1}(y/x) dy + f(x) = 2tan^{-1}(y/x) + f(x)$$

Since there is no term here that is just a function of y, we are safe to assume there will be no f(y) case in the x integral. But what if we do the x integral? Well...

$$ v = \frac{-2}{y}\int \frac{1}{1+y^2/x^2} dx = \frac{-2}{y}y tan^{-1}(x/y) dy = -2tan^{-1}(x/y) $$


We will note that these don't match identically... and with a quick graph we can see why.



(graph of x, with y chosen to be 1). 

Namely, in replacing the integraiton constant with the integration function, we forgot to note that the constant can still exist when the function does not. A simple shifting up and down will fix things. We can even see this in the full 3D plot.

Just to be extra sure that both functions are sufficient, we take their derivatives with the computer as a check.



And with all the derivatives matching, we just pick one and use it.

$$ v = 2tan^{-1}(y/x) $$

This is the object we're looking at. Since everything's going counterclockwise, we go from A to B to C then back to A on the integral, giving us three segments AB , BC, and CA. CA is going to be a purely real integral. 

Now this probably isn't going to work but we can at least try to see if Cauchy's Theorem applies here, demanding the loop be zero. AFter all, $z*$ is just the complex conjugate of whatever we put in, so there are no obvious discontinuities. Is it analytic, though? 

$z* = x - iy$ which means u=x and v=-y. However, the derivatives suggest 1=-1 which is not true, so this is not analytic and Cauchy's theorem does not apply. Unfortunate but expected. (That said the integral might still go to zero, as the complex conjugate is just a type of reflection.) 

If we write out a complex integral explicitly, we get:

$$ \int_\gamma f(z) dz = \int_\gamma u dx - \int_\gamma v dy + i\int_\gamma u dy + i\int_\gamma vdx $$

$$ = \int_a^b u \frac{dx}{dt} dt - \int_a^b v \frac{dy}{dt}dt + i\int_a^b u \frac{dy}{dt}dt + i\int_a^b v \frac{dx}{dt} dt $$

Now we need to define our paths carefully. Since this is purely geometric we can consider our "speed" through t to just be 1 unit per unit time. For the simplest case, let's do CA first.

$$ \int_C^A u \frac{dx}{dt} dt - \int_C^A v \frac{dy}{dt}dt + i\int_C^A u \frac{dy}{dt}dt + i\int_C^A v \frac{dx}{dt} dt $$

As y does not change at all those terms vanish. dx/dt is simply 1.

$$ = \int_C^A u dt + i\int_C^A v dt $$

now u=x and v=-y. In this path, v=0 everywhere, so that term vanishes as well. u=x, and x=t in this path, so...

$$ = \int_C^A t dt = \int_{-2}^0 x dx = 2 $$

The integral along CA is 2.

Naturally, that was the simple one. In the other two cases things will be different. First of all, t will not scale directly with x and y but rather with the length of the segment. Let's do AB next, which has a length of $\sqrt{2}$. Here, t ranges from 0 to $\sqrt{2}$ and the step where we just convert back to x in the end will not cut it. 

$$ \int_A^B u \frac{dx}{dt} dt - \int_A^B v \frac{dy}{dt}dt + i\int_A^B u \frac{dy}{dt}dt + i\int_A^B v \frac{dx}{dt} dt $$

First of all, u and v are x and -y, with the -y being due to the conjugate. 

Along this path, x and y are equal, since the line is at a 45 degree angle, which will simplify things. They both range from 0 to 1 wth t, specifically through x = y = $t/\sqrt{2}$, which would make the derivatives $1/\sqrt{2}$ in both cases.

All together this changes things to:

$$ = \int_A^B \frac{t}{\sqrt{2}} \frac{1}{\sqrt{2}} dt + \int_A^B \frac{t}{\sqrt{2}} \frac{1}{\sqrt{2}}dt + i\int_A^B \frac{t}{\sqrt{2}} \frac{1}{\sqrt{2}}dt - i\int_A^B \frac{t}{\sqrt{2}} \frac{1}{\sqrt{2}}dt $$

Note that the imaginary integrals completely cancel, and that the real integrals are the same. This becomes...

$$ = \int_0^\sqrt{2} t dt = 1$$

So right now we have 2+2=4. Will the other integral provide us -4? We shall find out.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

Let's see if Cauchy's Theorem applies. Is the funciton analytic? 

$$ \frac{z^2-1}{z}$$

Alas, no, we have a discontinuity at z=0, which we will need to cross over to make a full semicircle. This means we have a single pole. In order to evaluate this kind of contour we, instead of integrating a simple semicircle, will integrate a semicircle with a "hop" over the discontinuity. Now we have a contour that circles no discontinuities and thus has to go to zero. Let's consider each segment: let the semicircle we want to solve for be $\gamma$, the miniature semicircle be $\Gamma$, and the integrals over the line be L and R (for Left and Right)

$$ \int L + \int\Gamma + \int R + \int \gamma = 0 $$

Now, we want to solve for $\int \gamma$, so we need to find the contributions of all the others. 24.69 gives us the contribution of the infinitesimal semicircle, $-ia_{-1}\pi$. Now, $a_{-1}$ is the residue at the pole, and since we have a simple pole this can ve determined by

$$ a_{-1} = lim_{z\rightarrow 0} z f(z) = lim_{z\rightarrow 0} z^2 - 1 = -1 $$

This means the infinitesimal semicircle provides an integral of $ i\pi $.

As for the L and R integrals, they diverge as they approach 0. However, $\frac{x^2-1}{x}$ is perfectly symmetric (and odd), so the integral on the Left will always cancel out the integral on the Right! Thus, the line integral along the real axis contributes nothing, in the end, no matter how close we appraoch the discontinuity in the middle.

This means...

$$ \int \gamma + i\pi = 0 $$

$$ \Rightarrow \int_\gamma f(z) dz = -i\pi $$

And that is our integral.
