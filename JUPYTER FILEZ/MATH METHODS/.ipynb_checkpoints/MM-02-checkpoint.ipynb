# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$



Ironic note: we did every other part of the homework before doing this first part, since it seemed the least straightforward. So expect us to repeat definitions for hermitian and antihermitian a lot. 

To do this, we can use integration by parts. Noting that antihermitian means that the transpose conjugate is negative of the original matrix, we seek to show that the derivative operator D (which is effectively d/dt in this case) does exactly that to one of the functions. 

Integration by parts of the inner product gives us:

$$ \int_a^b f^*g dt = \left[ fDg \right]_a^b - \int_a^b F^*g' dt $$

Since g(a)=g(b)=0, the evaluated term goes to zero, leaving us with...

$$ \Rightarrow \int_a^b f^*g dt = - \int_a^b F^*Dg dt $$

Now we don't acutally know what F (the antiderivative of f) is, so we're going to apply the derivat8ive operator to both sides. 

$$ \Rightarrow D \int_a^b f^*g dt = - D \int_a^b F^*g' dt $$
$$ \Rightarrow \int_a^b (gDf^* + f^*Dg) dt = - \int_a^b f^*Dg + F^* g' dt $$

Mmm, this goes nowhere. 

Let's try applying the derivative operator to the inner product. 

$$ D \int_a^b f^* g dt = \int_a^b gDf^* dt + \int_a^b f^*Dg dt $$

Now hold on, we recognize that stecond term as part of integration by parts! $ab| = \int a'b + \int b'a$. WHcih means...

$$ = [f^*g]_a^b = 0$$

So the derivative operator acting on the inner product is zero. This tells us a useful relation:

$$ \int_a^b gDf^* dt + \int_a^b f^*Dg dt= 0$$
$$ \Rightarrow \int_a^b gDf^* dt = - \int_a^b f^*Dg dt $$

Since the integral on both sides is over the same bounds, it is the same operator acting on both and thus can be removed. 

$$ \Rightarrow gDf^* = - f^*Dg $$

Now, what if we suppose that f=g, that is, the operator is acting upon the inner product of a given vector with itself. Well, the relation must become...

$$ \Rightarrow gDg^* = - g^*Dg $$

Now this is true in a trivial case, if the derivative is zero, but ignoring that we have all we need right here. The only difference between g and $g^*$ is the conjugation, that is, the imaginary number flipping. Now, we need to be explicit with the ordering. Technically speaking, we need to write the term on the left as (Dg*)(g) since we were in an inner product, and the conjugate takes the left side. Now, (Dg*)(g), if we are to transpsoe 8it, would become (g)(Dg*). Then if we conjugate it, it becomes (g*)(Dg). Note that the sign hsa not flipped and yet we have the same form as the object on the left side. Thus, $gDg^*$ is its own transpose conjugate, so as a whole it is antihermitian. 

Since g is completley arbitrary, that means D must be antihermitian in order for this property to hold. 

If D is antihermitian, that is, its transpose conjugate is negative itself, it can be represtented as A+iB, where A and B are real matrices. iD would then equal iA-B. 

Now, for an antihermitian matrix, A has to be antisymmetric, in order for the real coefficients on one side to be the negative on the other. B has to be the reverse: symmetric, so that when the conjugate flip the sign it ends up negative. Thus we now have a situation where iD is not an antihermitian matrix, as -B+iA has the antihermitian properties in the wrong spot. What it DOES have is the properties of a **hermitian matrix**! B is symmetric, so the transpose does nothing, as it should. iA is antysymmetric so the transpose conjugate flips the sign twice, resulting in a net change of zero! And since hermitian operators are defined by being their own transpose conjugate, well, that means we have ourselves a hermitian matrix (operator).

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$



By definition, a unitary operator is one where it times its conjugate equals the identity, or $UU^*=U^*U=I$. The conjugate just flips the sign of i, so we can write this expression out in whole as:

$$ (A+iB)(A-iB) = (A-iB)(A+iB) = I $$
$$ \Rightarrow A^2 -iAB + iBA + B^2 = A^2 + iAB - iBA + B^2 = I $$
$$ \Rightarrow -iAB + iBA = + iAB - iBA = I - A^2 - B^2$$

We note that since the first two sections are equal to each other, AB and BA must commute. This is almost trivial. Anyway, since they commute, we can now let AB=BA.

$$ \Rightarrow -iAB + iAB = + iAB - iAB = I - A^2 - B^2$$
$$ \Rightarrow 0 = 0 = I - A^2 - B^2$$
$$ \Rightarrow A^2 + B^2 = I $$

Which is the other property we wanted to show. 

A hermitian operator is one that is equal to its own transpose conjugate. 

An antihermitian operator is one that is equal to the negative of its own transpose conjugate. 

Every operator (that can be expressed as a matrix) can be written as A+iB, where A and B are purely real. This also applies to hermitian and anti-hermitian operators. 

Let's declare our arbitrary matrix to be M, our hermitian to be H, and our antihermitian to be J (chosen rather arbitrarily). So if M=H+J, the entire thing means $A_M+iB_M = A_H+iB_H + A_J+iB_J$

Since the A and B matrices are purely real, we can split up the equality into two: $A_M=A_H+A_J$ and $iB_M=iB_H+iB_J \Rightarrow B_M=B_H+B_J$ 

Let's examine A first. The properties of $A_H$ are that the transpose terms are equal to each other. The property of $A_J$ is that the transpose terms are *opposite* each other. We are free to specify any of these terms as we wish with those criteria. Since we are freely able to define the diagonal without interfering with this, we can make any diagonal on the other side. As for the off diagonals, we just need to show that the result for two terms flipped across the diagonal can be freely defined. After all, all the pairs are independent of each other, so if we can show one we can show them all. 

Let the pair be (x,y). Here, (x,y) needs to be constructed from a combination of (a,a) and (b,-b), so x=a+b and y=a-b. We can solve this system of equations: a=x-b, y=x-2b, b=(x-y)/2, a=x - (x-y)/2. We could simplify, but we do not need to, a and b depend only on x and y, not each other, so there is always a solution. Thus, an arbitrary matrix can always be composed a transpose and "antitranspose" matrix. 

This is essentially a restatement of a more general propoperty, that every matrix can be composed of a symmetric and antisymmetric one. 

Now, we return to the B side of this proof. It turns out that the exact same steps apply, except now $B_H$ is the antisymmetric (inverse mirror) matrix, and $B_J$ the symmetric, since they were the coefficients of the complex portions that would have theri signs flip. All this would do is swap the a and b in our above proof.

Therefore, since both the real and imaginary parts can be freely specified to produce any component of an arbitrary matrix, any matrix operator can be written as a sum of Hermitian and Antihermitian operators. Though we will reiterate that this is just a special formulation of the "symmetric + antisymmetric" property of matrices.

We can write T itself as a matrix.

$$
\begin{bmatrix}
2 & -1 \\
1 & 2 \\
\end{bmatrix}
$$

The dagger can be written as "transpose conjugate." Since we have no imaginary components, all we do is transpose.

$$
\begin{bmatrix}
2 & 1 \\
-1 & 2 \\
\end{bmatrix}
$$

Which, when acting on the given vector, produces

$$\boxed{(2\alpha_1 + \alpha_2, -\alpha_1 + 2\alpha_2)} $$

which is our answer.

The commutation test is simple: Do they make the same matrix when multiplied in both orders?

$$ TT^\dagger =
\begin{bmatrix}
5 & 0 \\
0 & 5 \\
\end{bmatrix}
$$

$$ T^\dagger T =
\begin{bmatrix}
5 & 0 \\
0 & 5 \\
\end{bmatrix}
$$

So yes they do commute. 

Meanwhile the hermitian test is trivial. IS the matrix equal to its own transpose conjugate? **NO**. Thus it is not hermitian.
