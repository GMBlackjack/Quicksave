# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

First of all, this is an even function, so we can represent it with a pure cosine series. The range actually ensures that this will be perfectly periodic, though the boundaries at $\pm \pi$ are not guaranteed to be smooth. 

Finding the pure cosine series can actually be done by just looking at the positive half of the series. Here, we use the specific definition in Haberman's Applied Partial Differential Equations.

$$ f(t) = \sum_{n=0}^\infty A_n cos\frac{n\pi t}{L}  $$
$$ A_0 = \frac1L \int_0^L f(t) dt $$
$$ A_n = \frac2L \int_0^L f(t) cos\frac{n \pi t}{L} dt $$

Note that we are integrating specificlaly over the positive half, so f(t) might as well just be $sin(\omega t)$. Since we're integrating over t, the actual bounds are $[-\pi/\omega , \pi/\omega]$. 

$$ A_0 = \frac{1}{\pi/\omega} \int_0^{\pi/\omega} sin(\omega t) dt = \frac{1}{\pi/\omega} [-\frac{cos(\omega t)}{\omega}]_0^{\pi/\omega} = \frac{2}{\pi}$$

$$ A_n = \frac2L \int_0^L sin(\omega t) cos\frac{n \pi t}{L} dt $$
$$ = \frac2L \int_0^L \frac12 (sin(\omega t + \frac{n\pi t}{L}) + sin(\omega t - \frac{n\pi t}{L})) dt $$
$$ = \frac{1}{\pi /\omega} [-\frac{1}{\omega+\frac{n\pi}{\pi/\omega}}cos(\omega t + \frac{n\pi t}{\pi/\omega}) -\frac{1}{\omega-\frac{n\pi}{\pi/\omega}}cos(\omega t - \frac{n\pi t}{\pi/\omega})]_0^{\pi/\omega} $$


$$ = \frac{1}{\pi/\omega} [-\frac{1}{\omega+n\omega}cos(\pi+n\pi) -\frac{1}{\omega-n\omega}cos(\pi - n\pi)+\frac{1}{\omega+n\omega} +\frac{1}{\omega-n\omega}]$$

Since our bounds are very neat, we can replace the trig functions with alternating powers of (-1), and the plus or minus inside of them actually has no influence on the final product given the symmetry of cosine. n=0 makes them both 1, n=1 makes them both -1, and so on. Each one can just be $(-1)^n$ and all will be fine.

$$ = \frac{1}{\pi} [-\frac{1}{1+n}(-1)^n -\frac{1}{1-n}(-1)^n+\frac{1}{1+n} +\frac{1}{1-n}]$$

Notably n=0 will not actually be used since we had to find that separately. Also note that $\omega$ has completely divided out, which makes sence, as physically it is the frequency and the magnitude of each section in the series should not depend on it. Also, on even powers of n, everything goes to zero. On odd powers of n, the terms add.

$$ A_{even} = 0 $$
$$ A_{odd} = \frac{1}{\pi} [\frac{2}{1+n} +\frac{2}{1-n}]$$

However, this reveals a discontinuity. n=1 is a problem, for we're dividing by zero. We need to be more careful with the integral in that specific case. 

$$ A_1 = \frac2L \int_0^L sin(\omega t) cos\frac{\pi t}{L} dt $$
$$ = \frac2L \int_0^L \frac12 (sin(\omega t + \frac{\pi t}{L}) + sin(\omega t - \frac{\pi t}{L})) dt $$
Remember that $L = \pi/\omega$. This simplifies things to...

$$ = \frac{2}{\pi/\omega} \int_0^{\pi/\omega} \frac12 (sin(2\omega t) + sin(0)) dt $$
$$ = \frac{\omega}{\pi} \int_0^{\pi/\omega} sin(2\omega t) dt $$
$$ = \frac{\omega}{\pi} [-\frac{1}{2\omega}cos(2\omega t)]_0^{\pi/\omega}$$

Which goes to zero due to the factor of 2. So $A_1=0$

If we wanted to write it all out, we could say..

$$ f(x) = \frac{2}{\pi} + \sum_{n=3,5,7,...}^\infty \frac{1}{\pi}(\frac{2}{1+n}+\frac{2}{1-n}) $$

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

The extension in this case is -1-x, which satisfies f(-x) = -f(x).



And yet, curiously, to find the sine series we don't actually need to use the negative side, since the odd assumption is implicit in the form we use. 

$$ f(x) = \sum_{n=1}^\infty B_n sin\frac{n\pi x}{L} $$
$$B_n = \frac2L \int_0^L f(x) sin\frac{n \pi x}{L}$$

L in this case is just 1, so already things simplify markedly.

$$B_n = 2 \int_0^1 (1-x) sin(n \pi x)$$
$$ = 2 \left[ -\frac{1}{n\pi}cos(n \pi x) + \frac{1}{n^2\pi^2}sin(n \pi x) - \frac{x}{n \pi}cos(n \pi x) \right]_0^1 $$

$$ = 2 ( -\frac{1}{n\pi}cos(n \pi) - \frac{1}{n \pi}cos(n \pi) + \frac{1}{n\pi} ) $$
$$ = - \frac{4}{n \pi}cos(n \pi) + \frac{2}{n\pi} $$

Which we can split into even and odd results.

$$B_{nodd} = - \frac{6}{n \pi} $$
$$B_{neven} = - \frac{2}{n \pi} $$

And so...

$$ f(x) = \sum_{n=1,3,5...}^\infty -\frac{6}{n\pi} sin(n\pi x) + \sum_{n=2,4,6...}^\infty -\frac{2}{n\pi} sin(n \pi x) $$

The extension in this case is 1+x, which satisfies f(x)=f(-x)

Again, we only need to work with the positive side. Steps are basically the same as the previous problem. 

$$ f(x) = \sum_{n=0}^\infty A_n cos\frac{n\pi x}{L} $$
$$ A_0 = \frac1L \int_0^L f(x) dx $$
$$ A_n = \frac2L \int_0^L f(x) cos\frac{n \pi x}{L} dx $$

$$ A_0 = \int_0^1 (1-x) dx = 1 - \frac12 = \frac 12 $$

$$ A_n = 2 \int_0^1 (1-x) cos\frac(n \pi x) dx $$
$$ = 2 \left[ \frac{1}{n \pi}sin(n \pi x) - \frac{1}{n^2\pi^2}cos(n \pi x) - \frac{x}{n\pi}sin(n \pi x) \right]_0^1 $$
$$ = 2 (\frac{2}{n^2\pi^2}) $$
$$ = \frac{4}{n^2\pi^2} $$

$$ f(x) = \frac12 + \sum_{n=1}^\infty \frac{4}{n^2\pi^2} cos\frac(n \pi x) $$

The cosine series as it does not have any jump continuities, merely sharp angles.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

Now, while $x^2$ is a purely even function, we are not expanding $x^2$ We are expanding a funciton that repeats over $2\pi$. However, the Fourier series needs to espand around a point at the origin. Granted, we could move the origin, but by doing that the function is no longer even. What we have is a bunch of half-quadratics that start from zero and grow to a large number at $4\pi^2$ and then drop back down to zero immediately. So, for once, we need the full Fourier series.

$$ f(x) = A_0 + \sum_{n=1}^\infty A_n cos\frac{n \pi x}{L} + \sum_{n=1}^\infty B_n sin\frac{n \pi x}{L} $$
$$ A_0 = \frac{1}{2L} \int_{-L}^{L} f(x) dx $$
$$ A_n = \frac{1}{L} \int_{-L}^{L} f(x) cos\frac{n \pi x}{L}dx $$
$$ B_n = \frac{1}{L} \int_{-L}^{L} f(x) sin\frac{n \pi x}{L}dx $$

Now, we have two ways we can try to define this. We can keep the origin at x=0 and integrate over a piecewise function which would be $(x + 2\pi)^2$ or we move the origin to $x=\pi$. We choose the latter since it will allow us to integrate a continuous rather than piecewise function, even though this will change the funciton we integrate from $x^2$ to $(x+\pi)^2$ on the range $[-\pi,\pi]$.



$$ f(x) = A_0 + \sum_{n=1}^\infty A_n cos(n x) + \sum_{n=1}^\infty B_n sin(nx) $$
$$ A_0 = \frac{1}{2\pi} \int_{-\pi}^{\pi} (x+\pi)^2 dx $$
$$ A_n = \frac{1}{\pi} \int_{-\pi}^{\pi} (x+\pi)^2 cos(n x) dx $$
$$ B_n = \frac{1}{\pi} \int_{-\pi}^{\pi} (x+\pi)^2 sin(n x)dx $$

Now we could try u-substitution here... or we could jsut multiply out the binomial and get $x^2 + 2x\pi + \pi^2$ each term of which is simple enough to correspond with a known integral.

$$ A_0 = \frac{1}{2\pi} \int_{-\pi}^{\pi} x^2 + 2x\pi + \pi^2 dx =\frac{1}{2\pi} ( \frac{2\pi^3}{3} + 2\pi^3) = \frac{4\pi^2}{3} $$

$$ A_n = \frac{1}{\pi} \int_{-\pi}^{\pi} (x^2+2x\pi+\pi^2) cos(n x) dx $$
$$ = \frac{1}{\pi} \left[ \frac{x^2}{n}sin(nx) + \frac{2x}{n^2}cos(nx) -\frac{2}{n^3}sin(nx) + \frac{2\pi}{n^2}cos(nx) + \frac{2x\pi}{n} sin(nx) + \frac{\pi^2}{n}sin(nx) \right]_{-\pi}^{\pi}$$
$$ = \frac{1}{\pi} \left[ + \frac{2x}{n^2}cos(nx) + \frac{2\pi}{n^2}cos(nx) \right]_{-\pi}^{\pi}$$
$$ = \frac{1}{\pi} \left[ \frac{2x}{n^2}cos(nx)\right]_{-\pi}^{\pi}$$

$$ = \frac{1}{\pi} (\frac{2\pi}{n^2}(-1)^n - \frac{-2\pi}{n^2}(-1)^n)$$
$$ = (-1)^n\frac{1}{\pi} (\frac{4\pi}{n^2})$$
$$ = (-1)^n\frac{4}{n^2} $$

$$ B_n = \frac{1}{\pi} \int_{-\pi}^{\pi} (x^2+2x\pi+\pi^2) sin(n x) dx $$
$$ = \frac{1}{\pi} \left[ -\frac{x^2}{n}cos(nx) + \frac{2x}{n^2}sin(nx) +\frac{2}{n^3}cos(nx) + \frac{2\pi}{n^2}sin(nx) - \frac{2x\pi}{n} cos(nx) - \frac{\pi^2}{n}cos(nx) \right]_{-\pi}^{\pi}$$
$$ = \frac{1}{\pi} \left[ -\frac{x^2}{n}cos(nx) +\frac{2}{n^3}cos(nx) - \frac{2x\pi}{n} cos(nx) - \frac{\pi^2}{n}cos(nx) \right]_{-\pi}^{\pi}$$
$$ = \frac{1}{\pi} \left[ - \frac{2x\pi}{n} cos(nx) \right]_{-\pi}^{\pi}$$

$$ =  \left[ - \frac{2x}{n} cos(nx) \right]_{-\pi}^{\pi}$$
$$ =  - \frac{2\pi}{n}(-1)^n + \frac{-2\pi}{n}(-1)^n $$
$$ =  -(-1)^n\frac{4\pi}{n} $$

So we can write the entire series as:

$$ f(x) = \frac{4\pi^2}{3} + \sum_{n=1}^\infty (-1)^n\frac{4}{n^2} cos(n x) - \sum_{n=1}^\infty (-1)^n \frac{4\pi}{n} sin(nx) $$

Sanity check, actually graph this:



Of course, we *moved the origin* here. If we want a perfectly accurate reconstruction, we should move the origin back to where it originally was. This is a simple slide translation. 

$$ \boxed{f(x) = \frac{4\pi^2}{3} + \sum_{n=1}^\infty (-1)^n \frac{4}{n^2} cos(n (x+\pi)) - \sum_{n=1}^\infty (-1)^n \frac{4\pi}{n} sin(n(x+\pi))} $$

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

If we set x to 0, the Fourier series we found in **Problem 3** becomes

$$ f(0) = \frac{4\pi^2}{3} + \sum_{n=1}^\infty (-1)^n \frac{4}{n^2} + 0 $$

Now f(0) = $\pi^2$ due to the way we defined our origin. 

$$ \pi^2 = \frac{4\pi^2}{3} + \sum_{n=1}^\infty(-1)^n\frac{4}{n^2}  $$
$$ \Rightarrow -\frac{\pi^2}{3} = + 4 \sum_{n=1}^\infty(-1)^n\frac{1}{n^2}  $$
$$ \Rightarrow - \frac{\pi^2}{12} = \sum_{n=1}^\infty (-1)^n \frac{1}{n^2}  $$

The problem is that this is not the series we want (though wolfram does confirm that it is, indeed, what that series converges to). It does *contain* it, however:

$$ \Rightarrow -\frac{\pi^2}{12} = \sum_{n=1}^\infty \frac{1}{n^2} - 2\sum_{n=1,3,5...}^\infty \frac{1}{n^2}  $$

But that would require knowing what the odd version of the series is, and at that point one might as well know the total version. So, is there more information we can extract? We can try to use the bounds of the Fourier series, $\pm \pi$. However, there is a discontinuity here, so we have to be careful. The lower bound is 0, the upper bound is $4\pi^2$ so right in the middle is $2\pi^2$, which is what the Fourier series will approach at those points. The sine portion of the Fourier series still goes to zero, but now the cosine portion will start flip-flopping... 

$$ 2\pi^2 = \frac{4\pi^2}{3} + \sum_{n=1}^\infty (-1)^n \frac{4}{n^2} (-1)^n + 0 $$
$$ \Rightarrow \frac{2\pi^2}{3} = 4\sum_{n=1}^\infty\frac{1}{n^2} $$
$$ \Rightarrow \boxed{\frac{\pi^2}{6} = \sum_{n=1}^\infty\frac{1}{n^2}} $$

Which is the well-known result.
