PERTURBATION A

FIRST, find the exact energy levels. This actually isn't hard at all for this one, as we can rewrite the hamiltonian as:

$$ H = \frac{p^2}{2m} + \frac{1}{2}m(\omega^2+\alpha^2)x^2 $$

From which the energies are...

$$ E = (n+1/2)\hbar \omega \rightarrow (n+1/2)\hbar (\sqrt{\omega^2 + \alpha^2}) $$

Tah-dah! 

PERTURBATION B

The full hamiltonian is 

$$ H = \frac{p^2}{2m} + \frac{1}{2}m\omega^2 x^2 + bx $$

Sadly we can't just force this one into the known format, so we have to work out what the eigenenergies are normally. Writing out the Schrodinger equation:

$$ \left[ -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + \frac{1}{2}m\omega^2 x^2 + bx \right] \psi = E\psi $$
$$ \Rightarrow \left[ -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + \frac{1}{2}m\omega^2 x^2 + bx - E \right] \psi = 0 $$
$$ \Rightarrow \left[ \frac{\partial^2}{\partial x^2} - \frac{m^2\omega^2}{\hbar^2}x^2 - \frac{2mb}{\hbar^2}x + \frac{2mE}{\hbar^2} \right] \psi = 0 $$
