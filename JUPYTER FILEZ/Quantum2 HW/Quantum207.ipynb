## Problem 1



To start, we need to develop the potential function. Point hydrogen is $-e^2/r$. This term does not change for the exterior of the new model, for by Gauss' law we know that any symmetric potential distribution can be treated as a point charge so long as one is outside the charge itself, so this is accurate for rR. However, within the nucleus, the charge changes. We get our model for the central charge from [1], specifically **Example 2.7**. For inside the sphere, we have

$$ V(r) = \frac{1}{4\epsilon_0} \frac{q}{R} \rightarrow -\frac{e^2}{R} $$

We drop off the leading constant since it exists for all electric potentials and we've already left it off our model. We change q to $e^2$ because the potential given by [1] is an arbitrary one that is to be multiplied by the charge of the object in question: in this case the -e charge of the electron gets multiplied. 

Note that the potential inside the solid charged sphere is constant. This is because the electric field inside is zero, as it is with all uniformly charged materials.

Thus the potential becomes a piecewise function:

$$
V(r) = 
\begin{array}{cc}
  \{ & 
    \begin{array}{cc}
      -\frac{e^2}{r} & r  R  \\
      -\frac{e^2}{R} & r  R
    \end{array}
\end{array}
$$

Note how both say "or equal to" in their relations. This is because at the boundary r=R the two forms are exactly the same. 

We are asked to compute the correction as the expectation value of the new potential minus the normal Coulomb potential. Which means we actually use the following potential:

$$
V_-(r) = 
\begin{array}{cc}
  \{ & 
    \begin{array}{cc}
      0 & r  R  \\
      -\frac{e^2}{R}+\frac{e^2}{r} & r  R
    \end{array}
\end{array}
$$

Now we need the test function. This is given to us as the ground state function for the hydrogen atom, which is written out as:

$$ \psi_{100}(r,\theta,\phi) = \frac{1}{\sqrt{a_0^3}}e^{-r/a_0} $$

Fortunately for us this comes pre-normalized so we can skip that step. However, we are told not to vary any parameter here, so we treat everything as a known constant except r, which is the variable. Now, for the variational method, we find the expectation value of H for this wavefunction. H itself is, for our case

$$ H = \frac{p^2}{2m} - \frac{e^2}{R}+\frac{e^2}{r} = \frac{p^2}{2m} + e^2(\frac{1}{r}-\frac{1}{R}) $$

We note that this is only for *inside* the potential. Stating it piecewise:

$$
H(r) = 
\begin{array}{cc}
  \{ & 
    \begin{array}{cc}
      \frac{p^2}{2m} & r  R  \\
      \frac{p^2}{2m} + e^2(\frac{1}{r}-\frac{1}{R}) & r  R
    \end{array}
\end{array}
$$

Since we have a clear discontinuity within the wavefunction, we'll use the altered form for the Hamiltonian expectation.

$$ \langle H \rangle = \frac{\int\psi^*H\psi}{\int\psi^*\psi} = \frac{\int \frac{\hbar^2}{2m}\nabla \psi^*\nabla \psi +V\psi^*\psi}{\int\psi^*\psi} $$

We know that the function is normalized so the denominator in our case is 1, and we are only dealing with r so this is an integral over one dimension. (But we still have to add 4$r^2$ as it is still spherical coordinates). Furthermore, we write $ \nabla =  \frac{\partial}{\partial r} $ as the angular components are irrelevant. 

So we are finally able to set up

$$ \langle H \rangle = 4 \int [\frac{\hbar^2}{2m}\frac{\partial}{\partial r} \psi^*\frac{\partial}{\partial r} \psi +V\psi^*\psi] r^2 dr $$

Insert the wavefunction and take its derivatives.

$$ = 4 \int [\frac{\hbar^2}{2m}(\frac{-1}{a_0\sqrt{a_0^3}}e^{-r/a_0})^2 +V\frac{1}{a_0^3}e^{-2r/a_0}] r^2 dr $$

Now we introduce the potential V, which splits up the integral. This induces some annoyingly duplicate notation with e-charge and e-number, but context makes it clear which is which, for e-number is always an exponential. That said we will still differentiate the two via a subscript. 

$$ = 4 \int_0^R [\frac{\hbar^2}{2m}\frac{1}{a_0^5}e^{-2r/a_0} +(-\frac{e_c^2}{R}+\frac{e_c^2}{r})\frac{1}{a_0^3}e^{-2r/a_0}] r^2 dr + 4 \int_R^ [\frac{\hbar^2}{2m}\frac{1}{a_0^5}e^{-2r/a_0} +0] r^2 dr $$

$$ = \frac{4}{a_0^3} \left[ \int_0^R [\frac{\hbar^2}{2m}\frac{1}{a_0^2}e^{-2r/a_0} +(-\frac{e_c^2}{R}+\frac{e_c^2}{r})e^{-2r/a_0}] r^2 dr + \int_R^ [\frac{\hbar^2}{2m}\frac{1}{a_0^2}e^{-2r/a_0}] r^2 dr \right] $$

$$ = \frac{4}{a_0^3} \left[ \int_0^R \frac{\hbar^2}{2m}\frac{1}{a_0^2}r^2e^{-2r/a_0} -\frac{e_c^2}{R}r^2e^{-2r/a_0}+e_c^2re^{-2r/a_0}dr + \int_R^ \frac{\hbar^2}{2m}\frac{1}{a_0^2}r^2e^{-2r/a_0} dr \right] $$

$$ = \frac{4}{a_0^3} \left[ \int_0^R -\frac{e_c^2}{R}r^2e^{-2r/a_0}+e_c^2re^{-2r/a_0}dr + \int_0^ \frac{\hbar^2}{2m}\frac{1}{a_0^2}r^2e^{-2r/a_0} dr \right] $$

At this point integrals start being taken, calcualted using Geogebra. 

$$ = \frac{4}{a_0^3} \left[ \frac{e_c^2}{R}\left[ \frac{1}{8}e^{-2r/a_0}(4a_0r^2+4a_0^2r+2a_0^3) \right]_0^R - e_c^2\left[ \frac{1}{4}e^{-2r/a_0}(2a_0r+a_0^2) \right]_0^R - \frac{\hbar^2}{2m}\frac{1}{a_0^2}\left[ \frac{1}{8}e^{-2r/a_0}(4a_0r^2+4a_0^2r+2a_0^3) \right]_0^ \right] $$

Looking to be a negative energy at this point, which is exactly what it's supposed to be. Evaluate each segment invididually. 

$$ \left[ \frac{1}{8}e^{-2r/a_0}(4a_0r^2+4a_0^2r+2a_0^3) \right]_0^R 
= \frac{1}{8} \left( e^{-2R/a_0}(4a_0R^2+4a_0^2R+2a_0^3) - (2a_0^3) \right)
$$

$$ \left[ \frac{1}{4}e^{-2r/a_0}(2a_0r+a_0^2) \right]_0^R 
= \frac{1}{4} \left( e^{-2R/a_0}(2a_0R+a_0^2) - (a_0^2) \right)
$$

$$\left[ \frac{1}{8}e^{-2r/a_0}(4a_0r^2+4a_0^2r+2a_0^3) \right]_0^ 
= - (2a_0^3) 
$$

Now we insert back into our previous arrangement, simplifying with available constants.

$$ = \frac{4}{a_0^3} \left[ \frac{e_c^2}{8} \left( e^{-2R/a_0}(4a_0R+4a_0^2+\frac{2a_0^3}{R}) - \frac{2a_0^3}{R} \right) - \frac{e_c^2}{4} \left( e^{-2R/a_0}(2a_0R+a_0^2) - (a_0^2) \right) + \frac{\hbar^2a_0^2}{m}\right] $$

$$ =  \frac{e_c^2}{2a_0^3} \left( e^{-2R/a_0}(4a_0R+4a_0^2+\frac{2a_0^3}{R}) - \frac{2a_0^3}{R} \right) - \frac{e_c^2}{a_0^3} \left( e^{-2R/a_0}(2a_0R+a_0^2) - (a_0^2) \right) + \frac{\hbar^2}{ma_0} $$

This isn't exactly pleasant to look at, but it is our answer. Now, the question is, is this measurable? Well, taking R to be small, the exponential terms just vanish, leaving us with:

$$ =  \frac{e_c^2}{2a_0^3} \left(  - \frac{2a_0^3}{R} \right) - \frac{e_c^2}{a_0^3} \left( - (a_0^2) \right) + \frac{\hbar^2}{ma_0} $$

$$ =  -\frac{e_c^2}{R} + \frac{e_c^2}{a_0} + \frac{\hbar^2}{ma_0} $$

The 1/R term clearly dominates, as R << $a_0$. The remaining potential is negative, as it should be, and only requires us to plug in two values: fundamental charge and proton radius. (1.602e-19 C and 0.85e-15 m). This results in an energy of -3.019e-23 J. in eV, this is **0.000188 eV**. 

For reference, the ground state of hydrogen is -13.6 eV. This means the energy is most likely **not detectable**. This makes sense, as the nucleus is so tightly packed it might as well just be a point chargeatoms are mostly empty space, after all! 

PS: since we're second-guessing ourselves, we went ahead and calculated the value of the $\frac{\hbar^2}{ma_0}$ term, with the electron mass for m since the proton dominates the mass of the system. This produced an energy on the order of 1e-28 J, which is far far less than the 1/R term, so our leaving off of the other terms was justified. 

We mustn't forget doing this with a *muon*. The muon will not change the overall charge (it has -e just like the electron) but it *will* change the overall mass of the system. It would also change the initial wavefunctionspecifically by creating a new bohr radius $a_0$. 

Muon mass is 1.883e-28 kg. Proton is 1.673e-27 kg. They are comparable, neiher dominates, so the reduced mass becomes 1.683e-28. 

Now, how does the wave function change? $a_0$ is equivalent to $\frac{\hbar^2}{e^2}$ and we see a . This means we can scale from the old reduced mass (just the mass of the electron) and the new one. 

(This just explicitly means reduced mass. Not muon.) Sicne we have all these values, we calculate a new $a_0$ at 0.0027 angstroms, or 2.7e-13 m. This is somewhat close to the proton radius 0.85e-15m, so one might think the difference will be noticable. 

However, all the math we used above remains the same. Even with $a_0$ closer to R, the 1/R term still dominates in the final answer. The energy corrective will still be on the same order of magnitude as 0.000188 eV. 

However, what is the ground state energy for a muon and a proton? This would have the same formula as the energy for the ground state for the electron: $-\frac{\hbar^2}{2a_0^2}$. This value is normally 13.6eV for the electron, but we can scale it using our new  and bohr radius. This gives us a ground state energy of 0.00135 eV. This is on a comparable order of magnitude to 0.000188. So in this case, **yes the difference would be measurable.**

Which makes logical sense, as the muon is "orbiting" much closer in, where the size of the nucleus is no longer neglegible and things are not just empty space.
