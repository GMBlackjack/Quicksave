## Problem 1



First of all, we need to ask ourselves the precisde definition of the dimensionality of a group. The slides declare it to be the number of continuous, real parameters. 

The slides also give us the answers to all of these: GL(N,C)=$2N^2$, SL(N,C)=$2(N^2-1)$, U(N)=$N^2$, and SU(N)=$N^2-1$. The quesiton is *why*. 

Perhaps we need a more precise definition of a dimension. [1] states (in the case of vector fields) "If V (vector space) is a finitely generated F(field)-module, (i.e., has a finite basis) the cardinality of any basis is called the *dimension* of V and is denoted by $dim_FV$, or just dimV when F is clear from the context, and V is said to be *fininte dimensional* over F. If V is not finitely generated, V is said to be infinite dimensional (writen dimV=∞)"

The important thing to pick out here is "cardinality of the basis" which we can think of as "how many vectors/kets are required to get everything." 

From [1] we find that GL, the "general linear group," is written often as $GL_n(F)$ where n is the size of the matrices and F is the field of which they are a part. In the above notation, this is just GL(N), with n=N from the previous definition. Since we are composed of NxN matrices, we have NxN dimensions. (Don't we just have N? No, that would be for a specific type of matrix of a certain parameterization and pattern, this is the *general* group.) So the dimensions of GL(N) is $N^2$. We multiply this by two since we have the addition of the complex plane; now every dimension also has a potential imaginary component via z=a+bi. Thus, we have come to the correct answer of $2N^2$.

The SL(N) group is the special linear group which is parametrized by the determinant needing to be one, but is otherwise constructed the same way as GL(N). SL is clearly a subset of GL. But what is its dimensions? Well, if we set the determinant to 1, we are imposing a limitation on *one* value of the Group. We can easily imagine setting up an NxN matrix where adjusting one value of the matrix could adjust the determinant value (this is not always true, but it is at least possible in some cases). In this illustration, if we were to hold that value in the matrix fixed, the dimensionality would be reduced by one. In the complete case, we can deduce that a basis could be *constructed* for SL(N) that used one less parameter than GL(N). Say, if GL(N) had all its matrix spots filled but one, and that last one was filled in by adjusting the "Determinant" variable. Well, if we set the determinant variable, that last hole would be fored to be something. In every case, such a basis could be constructed. Thus, we can conclude that by limiting one number of freedom, we go from $N^2$ dimensions to $N^2-1$. Then we multiply by two when adding the complex addition to this, arriving at the correct answer of $2(N^2-1)$

U(N) is trivial at this point. It contains NxN matrices composed only of real numbers. Thus, its dimensionality is $N^2$, the correct answer. Curiously there is a limitation on the determinant of a Unitary matrix, however, it can still vary: ±1, which means there is still a parameter and not a constant here. If we force the determinant to be 1, we end up with SU(N), which has dimensionality $N^2-1$ by the same argument we had for SL(N). 

And with that the problem is done.

## Problem 2



Traditionally, the group $S_3$ is represented by {1, r, $r^2$, s, sr, s$r^2$} and is known as the Dihedral-6 group. (Since it is of order six). If stated in terms of our problem, this becomes {$360°, 120°, 240°,$ refleciton 1, reflection 3, reflection 2}. Yes, the numbering of the reflections is intentional, since if one were to go clockwise around, the reflections represented by sr and s$r^2$ would end up in reverse order. 
Note that a full 360 rotation is the identity in this case, so we've already found one of the four properties of a group. 

The MAIN property of a group is that it is closed, and it should be obvious that this is. Flipping a triangle along any axis of symmetry or rotating it any number of third turns will not make it anything other than a triangle composed of three points, it just shuffles the locations of the three poitns among the six representations. 

The third property of groups is inverse elements. Every element has an inverse, here they are paired:

(1,1)

(r,$r^2$)

($r^2$,r)

(s,s)

(sr,sr)

(s$r^2$,s$r^2$)

The reflection terms are their own inverse, as are the identities. Note that even though the reflections are composed of rotations that are not their own inverse, these rotations are being performed on "reflected" sides so they undo each other. Note that we explicitly state r and $r^2$ are inverses of each other; order sometimes matters in Groups, so each order of the operation needs to be considered.

The last property is that of association. That is, if one were to perform three operations in order (let's say sr, r, and s$r^2$ for example) that the parentheses could be drawn anywhere and the result would be the same. While an exhaustive proof of this is possible, it is annoying, so we shall simply justify it as true. In our example, without parentheses, we would perform a reflection, a rotation, and then a different reflection. Assuming our triangle strated in the identity state, it would end up in the rotated once state, or r. Now if we were to do it in another order, say, taking sr r first, which combine to form a simple s reflection, and then apply s to s$r^2$ we arrive at the rotated once state as well. Naturally this is just a demonstration and not a proof, so why does this occur?

If we think of a triangle as a sequence of numbered points 1,2,3 it should become evident. All these operations do is re-arrange the order of these elements in a way that makes sense as part of a triangle. A rotation turns 1,2,3 into 3,1,2 for instance. Now, it is always possible to apply a symmetry to an object that has already *been* rotated/reflected. in this case, applying a rotation on 3,1,2 again would make it 2,3,1. It also happens that rotating twice corresponds to $r^2$. The realization here is that every set of symmetry operations *also* maps directly to an arrangement of the three points on a triangle. The "s" and "r" representations could easily be replaced with actual pictures of the triangle in certain arrangements, which define both a method of rotation *and* a physical position. Due to this, we can combine any number of rotations in any order to arrive at a new "position" that can then be used as an "operation" on any other "position" we want to feed into it. Thus we can think of a string of operations (like r r r r r r r s) as divisible into two sections: the part that defines the position of the triangle, and the part that defines what to do with it. The thing is, we can draw the line for that anywhere, say, (rrrr)(rrrs) or (rrrrrrr)(s) and resolve each of them, it doesn't matter. We hope this justifies the associative property, as it is not as immediately obvious as the others. 

All this work makes it rather easy to construct a representation of what we want: we can represent the triangle as a vector of (1,2,3) and define an operation between them. Let us write down the positions of the various triangles and the operations they represent:

1 = (1,2,3)

r = (3,1,2)

$r^2$ = (2,3,1)

s = (1,3,2)

sr = (2,1,3)

s$r^2$ = (3,2,1)

We can define the operation between them rather simply: consider the coordinate in the first position to be in the "1" position. The operation acting upon another one takes the coordinate in the first position and moves it to where the "1" is located in the acting operator. For example: (3,1,2)(1,3,2) produces (2,1,3). Translated into our operations, this is r s = sr. It works! However, this is not strictly the type of representation the problem asks for: it wants $R^2$ or $R^3$ basis, which requires matrices and the operation being matrix multiplication. We select $R^3$ and quiclky identify the identity

$$ 1 \Rightarrow  \begin{bmatrix}
1 & 0 & 0 \\ 
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix} $$


This correlates *directly* to our previous model: it can be read as "1 in position 1, 2 in position 2, and 3 in position 3." Following this pattern, we can define the rest of the matrices.

$$ r \Rightarrow  \begin{bmatrix}
0 & 1 & 0 \\ 
0 & 0 & 1 \\
1 & 0 & 0
\end{bmatrix} $$

$$ r^2 \Rightarrow  \begin{bmatrix}
0 & 0 & 1 \\ 
1 & 0 & 0 \\
0 & 1 & 0
\end{bmatrix} $$

$$ s \Rightarrow  \begin{bmatrix}
1 & 0 & 0 \\ 
0 & 0 & 1 \\
0 & 1 & 0
\end{bmatrix} $$

$$ sr \Rightarrow  \begin{bmatrix}
0 & 1 & 0 \\ 
1 & 0 & 0 \\
0 & 0 & 1
\end{bmatrix} $$

$$ sr^2 \Rightarrow  \begin{bmatrix}
0 & 0 & 1 \\ 
0 & 1 & 0 \\
1 & 0 & 0
\end{bmatrix} $$

And this set of six 3x3 matrices fits what we wanted perfectly. All the s-matrices undo each other. The fascinating thing about the r matrix is that it physically "slides" the 1 values in the matrices to the right. Also curiously, one would have thought the backwards-diagonal matrix would be s, but it is acutally $sr^2$. Interesting.
