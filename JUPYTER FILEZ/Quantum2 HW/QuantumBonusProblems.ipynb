First potential: $k/r^2$

The classical differential cross section is ultimately given by

$$ \frac{d\sigma}{d\Omega} = \frac{b}{sin\Theta} \left| \frac{\partial \Theta}{\partial b}\right|^{-1} $$

The classical cross section itself is $\sigma$. The parameter b is the impact parameter, a perpendicula rdistance between the beam axis and incoming trajectory. $d\Omega$ is the solid angle of the detector, such that $d\Omega = 2 \pi sin \theta d\theta$. Lastly $\Theta$ is the final scattering angle. 

In essence all we really need to find here is $\Theta$ which is given by

$$ \Theta = \pi - 2 \int^\infty_{r_{min}} \frac{b}{r^2} \left[ 1-\frac{b^2}{r^2}-\frac{V(r)}{E} \right]^{-1/2} dr$$

The impact parameter b is our variable so we do not need to define it. E is just some constant. V(r) is provided by the problem as two different potentials. However, we do need to know $r_{min}$ but we can accomplish this using the interior of the integral:

$$ 1-\frac{b^2}{r^2}-\frac{V(r)}{E} = 0 $$

Which when solved produces $r_{min}$.

Substituting in our potential...

$$ 1 - \frac{b^2}{r^2} - \frac{k}{Er^2} = 0 $$
$$ \Rightarrow \frac{b^2}{r^2} + \frac{k/E}{r^2} = 1 $$
$$ \Rightarrow b^2 + \frac{k}{E} = r^2 $$
$$ \Rightarrow \sqrt{b^2 + \frac{k}{E}} = r $$

This is technically two solutions, both the positive and negative square root, but r needs to be positive so we leave it off. 

Now we work with $\Theta$ and solve as far as we can.

$$ \Theta = \pi - 2 \int^\infty_{r_{min}} \frac{b}{r^2} \left[ 1-\frac{b^2}{r^2}-\frac{k}{Er^2} \right]^{-1/2} dr$$
$$ = \pi - 2b \int^\infty_{r_{min}} \frac{1}{r^2} \frac{1}{\sqrt{1-\frac{b^2}{r^2}-\frac{k}{Er^2}}}  dr$$
$$ = \pi - 2b \int^\infty_{r_{min}} \frac{1}{r \sqrt{r^2-b^2-\frac{k}{E}}}  dr$$

This is in an integral form known, and even provided by the problem statement. 

$$ = \pi - 2b \left[ \frac{1}{\left| \sqrt{b^2+\frac{k}{E}} \right|} sec^{-1}\left( \frac{r}{ \sqrt{b^2+\frac{k}{E}}} \right)  \right]^\infty_{r_{min}}$$

This is zero at $r_{min}$. At infinity inverse sec goes to $\pi/2$ and so we end up with a result of

$$ = \pi - 2b \frac{1}{\left| \sqrt{b^2+\frac{k}{E}} \right|} \frac{\pi}{2} $$
$$ = \pi - \frac{b\pi}{\left| \sqrt{b^2+\frac{k}{E}} \right|} $$
$$ = \pi(1 - \frac{b}{\left| \sqrt{b^2+\frac{k}{E}} \right|}) $$

Which is about as simplified as we can get.

Our final answer is going to involve taking this derivative with respect to b. This derivative becomes...

$$ \frac{-\frac{k}{E}}{b^2\sqrt{b^2+\frac{k}{E}}+\frac{k}{E}\sqrt{b^2+\frac{k}{E}}} $$
$$ = \frac{-\frac{k}{E}}{(b^2+\frac{k}{E})\sqrt{b^2+\frac{k}{E}}} $$
$$ = \frac{-\frac{k}{E}}{(b^2+\frac{k}{E})^{3/2}} $$

So the differential cross section is then

$$ \frac{b}{sin\left( \pi(1 - \frac{b}{\left| \sqrt{b^2+\frac{k}{E}} \right|}) \right)} \frac{(b^2+\frac{k}{E})^{3/2}}{\frac{k}{E}} $$

POLARIZED SCATTERING OF SPIN, CROSS SECTIONS OF ALL POSSIBLE COMBINATIONS OF SPINS.

Okay so here's the deal, all states are a (spin)(space) state form, where each spin state and space state is eitehr symmetric or antisymmetric, and the combination of the two must match for the type of particle. In our case, we have BOSONS (integer spin), so everything must be symmetric. 

With two spin-1 particles we have the 2, 1, and 0 states. 

The 2 states first: |22> |21> |20> |2-1> |2-2>, for spin of course. 

These spin states are conglomerations of two particles: |11>|11> for |22>. This state is obviously symmetric as the interchange is irrelevant, and applying $J_-$ preserves symmetry so all the Spin-2 spin-states are symmetric, which means the space state must be symmetric as well.

The story is different for the |11> state, which is composed as a|11>|10> + b|10>|11>. This is antisymmetric as the interchange is visibly different AND it must be orthogonal to the |21> state. This means the space equation must be antisymmetric to cancel this out.

The last case is the |00> case, which will have a|11>|1-1> + b|10>|10> + c|1-1>|11> which is symmetric.  

The differential cross section would be of the form $D=(f^2 \pm f^2)^2$ (lecture 31 for full equation) where the plus occurs for antisymmetric space and minus for symmetric space. The interaction term is not here as it is independent of spin. This is, of course, the space equation.
