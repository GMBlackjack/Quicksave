First potential: $k/r^2$

The classical differential cross section is ultimately given by

$$ \frac{d\sigma}{d\Omega} = \frac{b}{sin\Theta} \left| \frac{\partial \Theta}{\partial b}\right|^{-1} $$

The classical cross section itself is $\sigma$. The parameter b is the impact parameter, a perpendicula rdistance between the beam axis and incoming trajectory. $d\Omega$ is the solid angle of the detector, such that $d\Omega = 2 \pi sin \theta d\theta$. Lastly $\Theta$ is the final scattering angle. 

In essence all we really need to find here is $\Theta$ which is given by

$$ \Theta = \pi - 2 \int^\infty_{r_{min}} \frac{b}{r^2} \left[ 1-\frac{b^2}{r^2}-\frac{V(r)}{E} \right]^{-1/2} dr$$

The impact parameter b is our variable so we do not need to define it. E is just some constant. V(r) is provided by the problem as two different potentials. However, we do need to know $r_{min}$ but we can accomplish this using the interior of the integral:

$$ 1-\frac{b^2}{r^2}-\frac{V(r)}{E} = 0 $$

Which when solved produces $r_{min}$.

Substituting in our potential...

$$ 1 - \frac{b^2}{r^2} - \frac{k}{Er^2} = 0 $$
$$ \Rightarrow \frac{b^2}{r^2} + \frac{k/E}{r^2} = 1 $$
$$ \Rightarrow b^2 + \frac{k}{E} = r^2 $$
$$ \Rightarrow \sqrt{b^2 + \frac{k}{E}} = r $$

This is technically two solutions, both the positive and negative square root, but r needs to be positive so we leave it off. 

Now we work with $\Theta$ and solve as far as we can.

$$ \Theta = \pi - 2 \int^\infty_{r_{min}} \frac{b}{r^2} \left[ 1-\frac{b^2}{r^2}-\frac{k}{Er^2} \right]^{-1/2} dr$$
$$ = \pi - 2b \int^\infty_{r_{min}} \frac{1}{r^2} \frac{1}{\sqrt{1-\frac{b^2}{r^2}-\frac{k}{Er^2}}}  dr$$
$$ = \pi - 2b \int^\infty_{r_{min}} \frac{1}{r \sqrt{r^2-b^2-\frac{k}{E}}}  dr$$

This is in an integral form known, and even provided by the problem statement. 

$$ = \pi - 2b \left[ \frac{1}{\left| \sqrt{b^2+\frac{k}{E}} \right|} sec^{-1}\left( \frac{r}{ \sqrt{b^2+\frac{k}{E}}} \right)  \right]^\infty_{r_{min}}$$

This is zero at $r_{min}$. At infinity inverse sec goes to $\pi/2$ and so we end up with a result of

$$ = \pi - 2b \frac{1}{\left| \sqrt{b^2+\frac{k}{E}} \right|} \frac{\pi}{2} $$
$$ = \pi - \frac{b\pi}{\left| \sqrt{b^2+\frac{k}{E}} \right|} $$
$$ = \pi(1 - \frac{b}{\left| \sqrt{b^2+\frac{k}{E}} \right|}) $$

Which is about as simplified as we can get.

Our final answer is going to involve taking this derivative with respect to b. This derivative becomes...

$$ \frac{-\frac{k}{E}}{b^2\sqrt{b^2+\frac{k}{E}}+\frac{k}{E}\sqrt{b^2+\frac{k}{E}}} $$
$$ = \frac{-\frac{k}{E}}{(b^2+\frac{k}{E})\sqrt{b^2+\frac{k}{E}}} $$
$$ = \frac{-\frac{k}{E}}{(b^2+\frac{k}{E})^{3/2}} $$

So the differential cross section is then

$$ \frac{b}{sin\left( \pi(1 - \frac{b}{\left| \sqrt{b^2+\frac{k}{E}} \right|}) \right)} \frac{(b^2+\frac{k}{E})^{3/2}}{\frac{k}{E}} $$

Second potential: $k/r$

$r_{min}$ is different for this one:

$$ 1 - \frac{b^2}{r} - \frac{k}{Er} = 0 $$
$$ \Rightarrow \frac{b^2}{r} + \frac{k/E}{r} = 1 $$
$$ \Rightarrow b^2 + \frac{k}{E} = r $$

Now we work with $\Theta$ and solve as far as we can.

$$ \Theta = \pi - 2 \int^\infty_{r_{min}} \frac{b}{r^2} \left[ 1-\frac{b^2}{r^2}-\frac{k}{Er} \right]^{-1/2} dr$$
$$ = \pi - 2b \int^\infty_{r_{min}} \frac{1}{r^2} \frac{1}{\sqrt{1-\frac{b^2}{r^2}-\frac{k}{Er}}}  dr$$
$$ = \pi - 2b \int^\infty_{r_{min}} \frac{1}{r \sqrt{r^2-b^2-\frac{kr}{E}}}  dr$$

Integrate with the provided integral.

$$ = \pi - 2b \left[ \frac{1}{b} sin^{-1} \left( \frac{-\frac{kr}{E}-2b^2}{r\sqrt{\frac{k^2}{E^2}+4b^2}} \right) \right]^\infty_{r_min} $$

$$ = \pi - 2 \left[ sin^{-1} \left( \frac{-\frac{k}{E}}{\sqrt{\frac{k^2}{E^2}+4b^2}} \right) -sin^{-1} \left( \frac{-\frac{kb^2}{E}-\frac{k^2}{E^2}-2b^2}{(b^2 + \frac{k}{E})\sqrt{\frac{k^2}{E^2}+4b^2}} \right) \right] $$

Which, while ugly, can still be substituted in. Note that it does not depend on r or theta anymore, but it does still depend on b. And it'll be an annoying derivative to take... but luckily we have computers for that. [1] The derivative with respect to b is...

$$ -4 \frac{k}{E} \frac{b}{|b| (\frac{k^2}{E^2}+4b^2)}- \frac{8b^3((\frac{k}{E}+2)b^2+2\frac{k^2}{E^2}-2\frac{k}{E})}{(b^2+\frac{k}{E})^2(4b^2+\frac{k^2}{E^2})^{3/2}\sqrt{1-\frac{(-\frac{k}{E}b^2-2b^2-\frac{k^2}{E^2})^2}{(b^2+\frac{k}{E})^2(4b^2+\frac{k^2}{E^2})}}} $$

Which is immensely ugly but we can still turn it into our result. 

$$ \frac{b}{sin(\pi - 2 \left[ sin^{-1} \left( \frac{-\frac{k}{E}}{\sqrt{\frac{k^2}{E^2}+4b^2}} \right) -sin^{-1} \left( \frac{-\frac{kb^2}{E}-\frac{k^2}{E^2}-2b^2}{(b^2 + \frac{k}{E})\sqrt{\frac{k^2}{E^2}+4b^2}} \right) \right])} \left( -4 \frac{k}{E} \frac{b}{|b| (\frac{k^2}{E^2}+4b^2)}- \frac{8b^3((\frac{k}{E}+2)b^2+2\frac{k^2}{E^2}-2\frac{k}{E})}{(b^2+\frac{k}{E})^2(4b^2+\frac{k^2}{E^2})^{3/2}\sqrt{1-\frac{(-\frac{k}{E}b^2-2b^2-\frac{k^2}{E^2})^2}{(b^2+\frac{k}{E})^2(4b^2+\frac{k^2}{E^2})}}} \right)^{-1} $$

There is no doubt a way to simplify this, but it's technically correct. At least it should be.

This is actually two questions. 

The scattering wave packet is

$$\Psi^\pm(\vec{r},t) = \int \frac{1}{(2\pi)^{3/2}} \phi (\vec{k}-\vec{k_0}) e^{-i\vec{k}\cdot\vec{r_0}} \psi^\pm_\vec{k} (\vec{r}) e^{-i\hbar k^2t/2m} d^3k$$

And if we apply approximations--WAIT no we're not allowed to apply approximations! The next equation is not right:

$$\Psi^\pm(\vec{r},t) \approx e^{-i\vec{k_0}\cdot\vec{r_0}} e^{-i\hbar k_0^2t/2m} \int \phi(\vec{k'})e^{-i\vec{k'}\cdot(\vec{r_0}+\hbar\vec{k_0}t/m)} \psi^\pm_{\vec{k_0}+\vec{k'}}(\vec{r}) d^3k' $$

Howere, it is true that the, the time-dependent Schrodinger equation is 

$$ i\hbar \frac{\partial}{\partial t} \Psi = H \Psi  $$

If we insert our wavefunction in, we note that it is an eigenfunction of H and can adjust:

$$ i\hbar \frac{\partial}{\partial t} \Psi^\pm = \frac{\hbar^2k^2}{2m} \Psi^\pm$$

But we are NOT just using the approximation, so the next step is simultaneously more difficult and easier. See, the only t is inside the integral, but its in an exponential--exponentials remain the sive when integreated or differnetiated (which means we better not be integrating over a range, otherwise the k would vanish.) Which means that, whatever form it takes, the exponetnial remains, so we know the constant next to the t is what we take out. In other terms...

$$ i\hbar \frac{-i\hbar k^2}{2m} \Psi^\pm = \frac{\hbar^2k^2}{2m} \Psi^\pm $$
$$ \Rightarrow \frac{\hbar^2 k^2}{2m} \Psi^\pm = \frac{\hbar^2k^2}{2m} \Psi^\pm $$

Which is all we sought to show.

NOTE: we actually cannot evaluate the integral, all of this is done INSIDE the integrand!

To work with this we'd need to find out what the incident and scattered wavefunctions are, at least in terms of $\Psi$. Let us work with the incident case first. From the slides, we deduce:

$$ \Psi_{inc} = e^{\frac{i\hbar k_0^2 t}{2m}} \Psi(\vec{r}-\frac{\hbar \vec{k_0}}{m}t,0) $$

The full current relation is:

$$ \vec{j_{inc}} = \frac{\hbar}{2mi} \left[ e^{\frac{-i\hbar k_0^2 t}{2m}}\Psi^* \vec\nabla e^{\frac{i\hbar k_0^2 t}{2m}}\Psi - e^{\frac{i\hbar k_0^2 t}{2m}}\Psi \vec\nabla e^{\frac{-i\hbar k_0^2 t}{2m}}\Psi^* \right]$$
$$ = \frac{\hbar}{2mi} \left[ \Psi^* i\vec{k_0} \Psi + \Psi i\vec{k_0} \Psi^* \right]$$
$$ = \frac{\hbar \vec{k_0}}{m} \left[ \Psi^* \Psi \right]$$
$$ = \frac{\hbar \vec{k_0}}{m} \left| \Psi \right|^2$$

Which is the condensed form of what we set out to show. 

SECOND QUESTION: Derive the final forms of the incident and scattered currents. 

The starting forms of the currents are
$$ \vec{j_{inc}} = \frac{\hbar}{2mi} \left[ \Psi_{inc}^* \vec\nabla \Psi_{inc} - c.c. \right]$$
$$ \vec{j_{sc}} = \frac{\hbar}{2mi} \left[ \Psi_{sc}^* \vec\nabla \Psi_{sc} - c.c. \right]$$

And the final forms are:
$$ \vec{j_{inc}} \approx \frac{\hbar \vec{k_0}}{m} \left| \Psi(\vec{r}-\frac{\hbar \vec{k_0}}{m}t,0) \right|^2$$
$$ \vec{j_{sc}} \approx \frac{\hbar \vec{k_0}}{m} \hat{r} \frac{1}{r^2} |f^+_{\vec{k_0}}(\vec{\hat{r}})|^2 \left| \Psi(\vec{r}-\frac{\hbar \vec{k_0}}{m}t,0) \right|^2$$


Now to do the same with the other one. 

$$ \Psi_{sc} = e^{\frac{i\hbar k_0^2 t}{2m}}  f_{\vec{k_0}}(\vec{r}) \frac{1}{r} \Psi(\hat{k_0}r-\frac{\hbar \vec{k_0}}{m}t,0) $$

The full current relation is:

$$ \vec{j_{sc}} = \frac{\hbar}{2mi} \left[ e^{\frac{-i\hbar k_0^2 t}{2m}}f^*_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi^* \vec\nabla e^{\frac{i\hbar k_0^2 t}{2m}}f_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi - e^{\frac{i\hbar k_0^2 t}{2m}}f_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi \vec\nabla e^{\frac{-i\hbar k_0^2 t}{2m}}f^*_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi^* \right]$$
$$ = \frac{\hbar}{2mi} \left[ f^*_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi^* i\vec{k_0}\hat{r} f_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi + f_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi i\vec{k_0}\hat{r} f^*_{\vec{k_0}}(\vec{r}) \frac{1}{r}\Psi^* \right]$$
$$ = \frac{\hbar \vec{k_0}\hat{r}}{mr^2} \left[ \Psi^* \Psi f_{\vec{k_0}} f^*_{\vec{k_0}} \right]$$
$$ = \frac{\hbar \vec{k_0}\hat{r}}{mr^2} \left| \Psi \right|^2 \left| f_{\vec{k_0}} \right|^2 $$

Which is the condensed form of what we set out to show.
