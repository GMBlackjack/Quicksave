## Problem 1



First of all, we will use the model for perpendicular acceleration suggested in the slides: $\delta \vec\beta = \delta\beta\hat{y}$. 

Right away we don't even need to worry about the x coordinate—x''=x'=x because we have no components in the x-direction, just y and z (and this is why we defined the perpedicular acceleration in the y direction). However, all the other directions aren't so simple. 

Remember, any time we get a $\delta\beta^2$ we ignore it as "small" due to us using the first order approximation. Before we begin, though, we need to examine a good approximation for $\tilde\gamma$

$$ \tilde\gamma = \frac{1}{\sqrt{1-|\beta\hat{z} + \delta\beta\hat{y}|^2}} $$
$$ = \frac{1}{\sqrt{1-\beta^2 - \delta\beta^2}} $$

The squared term is small, which reduces this to:

$$ = \frac{1}{\sqrt{1-\beta^2}}  = \gamma$$  

Which simplifies calculations considerably. The coordinate relations for t', z', and y', remain the same, as $\vec{\delta\beta}$ does not factor into their calculation at all. 

Now that that's over with, let's start with t''. 

$$ t'' = \tilde\gamma[t-(\beta\hat{z}+\delta\beta\hat{y})\cdot\vec{r}] $$

The dot product quickly lets us get rid of the directionalities, and we can substitute in $\gamma$

$$ = \gamma[t -\beta z- \delta\beta y] $$

t, z, and y can all be expanded. 

$$ t'' = \gamma[\gamma(t'+\beta z') -\beta \gamma(z'+\beta t')- \delta\beta y'] $$
$$ = \gamma^2t'+\gamma^2\beta z' -\beta \gamma^2z' - \beta^2 \gamma^2 t'- \gamma\delta\beta y' $$

Remember $\gamma = \frac{1}{\sqrt{1-\beta^2}}$

$$ = \gamma^2 (1-\beta^2) t' - \gamma\delta\beta y' $$
$$ = t' - \gamma\delta\beta y' $$

**this is the correct relation for t''.**


We need to examine the physical coordinates now, xyz, all summed up in $\vec{r}$.

$$\vec{r''} = \vec{r} + \frac{\tilde\gamma-1}{|\beta\hat{z}+\delta\beta\hat{y}|^2}[(\beta\hat{z}+\delta\beta\hat{y})\cdot\vec{r}](\beta\hat{z}+\delta\beta\hat{y}) - \tilde\gamma(\beta\hat{z} + \delta\beta\hat{y})t$$

$$\vec{r''} = \vec{r} + \frac{\tilde\gamma-1}{\beta^2+(\delta\beta)^2}[\beta z+\delta\beta y](\beta\hat{z}+\delta\beta\hat{y}) - \tilde\gamma(\beta\hat{z} + \delta\beta\hat{y})t$$

This quckly confirms that x''=x as we already knew. However, it also gives us:


$$ y'' = y + \frac{\gamma-1}{\beta^2+(\delta\beta)^2}[\beta\delta\beta z+(\delta\beta)^2 y] - \gamma \delta\beta t$$

$$ z'' = z + \frac{\gamma-1}{\beta^2+(\delta\beta)^2}[\beta^2 z+ \beta\delta\beta y] - \gamma\beta t$$

All $(\delta\beta)^2$ terms vanish. 

$$ y'' = y + \frac{\gamma-1}{\beta^2}\beta\delta\beta z - \gamma \delta\beta t$$

$$ z'' = z + \frac{\gamma-1}{\beta^2}[\beta^2 z+ \beta\delta\beta y] - \gamma\beta t$$

Which match what are shown on the slides for this step. Now we examine one at a time and get it in terms of primes. Since x already known, y is next.

$$ y'' = y' + \frac{\gamma-1}{\beta^2}\beta\delta\beta \gamma(z'+\beta t') - \gamma \delta\beta \gamma (t'+\beta z')$$
$$ = y' + \frac{\gamma-1}{\beta^2}\beta\delta\beta\gamma z'+ \frac{\gamma-1}{\beta^2}\beta^2\delta\beta\gamma t' - \gamma^2\delta\beta t'- \gamma^2\beta\delta\beta z'$$
$$ = y' + \frac{\gamma-1}{\beta}\delta\beta\gamma z'+ (\gamma-1)\delta\beta\gamma t' - \gamma^2\delta\beta t'- \gamma^2\beta\delta\beta z'$$
$$ = y' + \frac{\gamma}{\beta}\delta\beta\gamma z' - \frac{1}{\beta}\delta\beta\gamma z' + \delta\beta\gamma^2 t' - \delta\beta\gamma t' - \gamma^2\delta\beta t'- \gamma^2\beta\delta\beta z'$$
$$ = y' - \gamma\delta\beta t'+ \frac{\gamma}{\beta}\delta\beta\gamma z' - \frac{1}{\beta}\delta\beta\gamma z' - \gamma^2\beta\delta\beta z'$$
$$ = y' - \gamma\delta\beta t'+ \frac{1}{\beta} \left( \gamma^2\delta\beta z' - \delta\beta\gamma z' - \gamma^2\beta^2\delta\beta z' \right)$$
$$ = y' - \gamma\delta\beta t'+ \frac{1}{\beta} \left( \gamma^2(1-\beta^2)\delta\beta z' - \delta\beta\gamma z' \right)$$
$$ = y' - \gamma\delta\beta t'+ \frac{1}{\beta} \left( \delta\beta z' - \delta\beta\gamma z' \right)$$
$$ = y' - \gamma\delta\beta t'- \frac{1}{\beta} (\gamma - 1) \delta\beta z'$$

Which is **the correct relation for y''**

$$ z'' = \gamma(z' + \beta t') + \frac{\gamma-1}{\beta^2}[\beta^2 \gamma(z' +\beta t') + \beta\delta\beta y'] - \gamma\beta \gamma(t' + \beta z')$$
$$ = \gamma z' + \gamma\beta t' + \frac{\gamma-1}{\beta^2}[\beta^2 \gamma z' +\gamma\beta^3 t' + \beta\delta\beta y'] - \gamma^2\beta t' - \gamma^2\beta^2 z'$$
$$ = \gamma z' + \gamma\beta t' + (\gamma-1)[\gamma z' +\gamma\beta t' + \frac{1}{\beta}\delta\beta y'] - \gamma^2\beta t' - \gamma^2\beta^2 z'$$
$$ = \gamma z' + \gamma\beta t' + \gamma^2 z' +\gamma^2\beta t' + \frac{\gamma}{\beta}\delta\beta y' - \gamma z' - \gamma\beta t' - \frac{1}{\beta}\delta\beta y' - \gamma^2\beta t' - \gamma^2\beta^2 z'$$
$$ = \gamma^2 z' + \frac{\gamma}{\beta}\delta\beta y' - \frac{1}{\beta}\delta\beta y' - \gamma^2\beta^2 z'$$
$$ = \gamma^2 (1-\beta^2) z' + \frac{\gamma-1}{\beta}\delta\beta y'$$
$$ = z' + \frac{\gamma-1}{\beta}\delta\beta y'$$

Which is **the correct relation for z''**

## Problem 2



Since degeneracy can be ignored we can use regular perturbaiton theory here, that is:

$$ E_{nl,rel}^{(1)} = \langle nlm | V_{rel} | nlm \rangle $$

Where

$$ V_{rel} = -\frac{p^4}{8m^3c^2} $$

Now the trick to evaluating this is splititng $p^4$ into $p^2p^2$ which have known eigenvalues.

$$ E_{nl,rel}^{(1)} = -\langle nlm | p^2\frac{1}{8m^3c^2} p^2 | nlm \rangle $$
$$ = -\langle nlm | (2m(E_n-V_N))\frac{1}{8m^3c^2}(2m(E_n-V_N)) | nlm \rangle $$
$$ = -\frac{1}{2mc^2}\langle nlm | (E_n-V_N)^2 | nlm \rangle $$
$$ = -\frac{1}{2mc^2}\langle nlm | E_n^2-2E_nV_N+V_N^2 | nlm \rangle $$
$$ = -\frac{E_n^2}{2mc^2}\langle nlm | 1-2\frac{V_N}{E_n}+\frac{V_N^2}{E_n^2} | nlm \rangle $$
$$ = -\frac{E_n^2}{2mc^2} \left[ 1 - \langle nlm | 2\frac{V_N}{E_n} | nlm \rangle + \langle nlm |\frac{V_N^2}{E_n^2} | nlm \rangle \right]$$
$$ = -\frac{E_n^2}{2mc^2} \left[ 1 - \frac{2}{E_n}\langle nlm | V_N | nlm \rangle + \frac{1}{E_n^2}\langle nlm |V_N^2 | nlm \rangle \right]$$


(A note on irony: after we had finmished deriving this we found that it was just equation 5.101.)

Which brings us to the fact that we really need to know what $V_N$ is. It is the electric potential energy due to the nucleus, or $-Ze^2/r$. If we take it for hydrogen Z=1 and the potential is $-e^2/r$. 

$$ = -\frac{E_n^2}{2mc^2} \left[ 1 + \frac{2Ze^2}{E_n}\langle nlm | \frac{1}{r} | nlm \rangle + \frac{Z^2e^4}{E_n^2}\langle nlm |\frac{1}{r^2}| nlm \rangle \right]$$


The problem now is that our wavefuncitons |nlm> are not exactly neat to take integrals over, especially since we're doing it arbitrarily for all nlm. 

It is worth noting that $E_n = -\frac{me^4}{2\hbar^2n^2}$. Equations 5.102 and 5.103 provide the answers for the unknown expectation values in our expression, but work still needs to be shown.

First, the 1/r case. There appear to be multiple methods for calculating this, but we choose to do the one suggested by Griffiths: applying the three dimensional Virial theorem. 

From Griffiths 3.113 we have

$$2\langle T \rangle = \langle x \frac{dV}{dx} \rangle$$

Which is easily converted into the 3D version. 

$$2\langle T \rangle = \langle \vec{r} \cdot \nabla V \rangle$$

Rather than trying to evaluate via symmetric potentials, we can actually show what $\vec{r} \cdot \nabla V$ equals.

$$ (x,y,z) \cdot (-Ze^2) \left(\frac{-x}{r^3} ,\frac{-y}{r^3} ,\frac{-z}{r^3}  \right)  $$
$$ = Ze^2 \frac{x^2+y^2+z^2}{r^3}  $$
$$ = Ze^2 \frac{1}{r}  $$

Which is exactly what we started with aside from a sign, which means

$$2\langle T \rangle = -\langle V \rangle$$

The kinetic energy term can just be throught of as the unperturbed Hamiltonian. Which, while we think this is obvious, the sign is uncertain, so we derive it manually:

$$ \langle T \rangle = \langle nlm | \frac{{p^2}}{2m} | nlm \rangle $$
$$ = \langle nlm | E_n - V_N | nlm \rangle $$
$$ =  E_n - \langle V_N \rangle $$

Which means we now have two eqiations, two unknowns. Solving simultaneously for the expectation of T and $V_N$ we get $-E_n$ and $2E_n$ simultaneously. This means we don't even have to work out the 1/r case like we thought. 

This means our result is currently:

$$ = -\frac{E_n^2}{2mc^2} \left[ 1 -4  + \frac{Z^2e^4}{E_n^2}\langle nlm |\frac{1}{r^2}| nlm \rangle \right]$$

$$ = -\frac{E_n^2}{2mc^2} \left[ -3 + \frac{Z^2e^4}{E_n^2}\langle nlm |\frac{1}{r^2}| nlm \rangle \right]$$


Now for the $1/r^2$ form. No Virial Theorem to save us this time. 

The textbook suggests solving a perturbation to the hydrogenic atom directly, just with an added $\gamma^2/r^2$ term. Rather than try to solve hte radial wave equation again, let's see if we can make its pattern match the normal hydrogenic solution. 

The nomral hydrogenic modified radial wave equation is:

$$ \left[\frac{d^2}{dr^2}  + \frac{2m}{\hbar^2}E + \frac{2\mu Ze^2}{\hbar^2r} - \frac{l(l+1)}{r^2}\right]u=0$$

Where the energy equals $ \frac{Z^2m e^4}{2\hbar^2}$. 

Of course this is just the 3energy scale, the actual energy get n-dependence added: $E_n = -E \frac{1}{n^2} = -\frac{Z^2me^4}{2 \hbar^2 n^2}$ 

Now if we add the suggested perturbaiton, the modified radial wave equation becomes:

$$ \left[\frac{d^2}{dr^2}  + \frac{2m}{\hbar^2}E + \frac{2\mu Ze^2}{\hbar^2r} - \frac{l(l+1) - \gamma}{r^2}\right]u=0$$

We note that $\gamma = 2mZ^2e^4/\hbar^2$ which we deduce from the position of the normal coulomb term, $Ze^2$. Anyway, this will give us a new l' value based on the solution to this equation:

$$ l(l+1) - \gamma = l'(l'+1) $$

Which we solve using Wolfram [1] . This becomes

$$ l' =  \frac{1}{2} \left( \pm \sqrt{(2l+1)^2-4\gamma} -1 \right) $$

Naturally we want the positive version as when $\gamma=0, l=l'.$



Anyway we now have an l' quantum number that definitely doesn't just take integer values. We can now solve as though we were doing the normal problem, by comparing values of $\rho_0$

$$ \rho_0 = \sqrt{\frac{2m}{-E}}\frac{Ze^2}{\hbar} $$

and

$$ \rho_0 = 2(N+l+1) $$

However, any existences of l now equal l'. This eventually leads us to...

$$ E = -\frac{mZ^2e^4}{2\hbar^2(N+l'+1)^2} $$

Note that N+l+1 = n by definition. So if we can put l' in terms of some l+c, we can get our quantum number n back into it. 

$$ N+l'+1 = N + \frac{1}{2}\sqrt{(2l+1)^2-4\gamma}+\frac{1}{2} $$

Now for small $\gamma$ this reverts to

$$ = N + \frac{1}{2}(2l+1) +\frac{1}{2} - \frac{\gamma}{2l+1} = N + l + 1 - \frac{\gamma}{2l+1} $$
$$ = n - \frac{\gamma}{2l+1} $$

But since $\gamma = 2mZ^2e^4/\hbar^2$ can we really cannot be sure this is small compared to everything?

If we expand E around $\gamma=0$, in both cases the first term is the normal expected energy. So the first order correction is...

$$ \frac{mZ^2e^4}{\hbar^2(2l+1)n^3}\gamma = \frac{mZ^2e^4}{\hbar^2(2l+1)n^3}\frac{2mZ^2e^4}{\hbar^2} $$

If we forcibly substitute in $E_n^2$ we arrive at...

$$ \frac{8n}{(2l+1)}E_n^2 $$

This gives us our final answer:

$$ E^{(1)}_{nl,rel}= -\frac{E_n^2}{2mc^2} \left[ -3 + \frac{4n}{l+1/2} \right]$$

## Problem 3



To begin, let us define $H_{00} = \frac{p^2}{2m} + V_n(r)$. Let |nlms> be our wavefunction, which can split into radial and angular components that are independent of each other (in terms of variables) as how the normal hydrogen wavefunction is, in the form RY. 

From the problem, we know $H_{00}|nlms> = e_n|nlms>$. Due to the independence of radial and angular parts, this will even be true for the adjusted wavefunction with the perturbation--in botht he exact and first-order correction. Now, we haven't the foggiest idea of the exact form of |nlms> since we have no idea what $H_{00}$ actually is, but we know it's only r-dependant. 

Now, it would help if we could restate that dot product.

$$ \vec{J} = \vec{L} + \vec{S} $$
$$ \Rightarrow J^2 = L^2 + S^2 + 2 \vec{L} \cdot \vec{S} $$
$$ \Rightarrow 2 \vec{L} \cdot \vec{S} = \frac{J^2-L^2-S^2}{2} $$

Which matches what is shown on the slides.  

It will help to get all our eigenvalues sorted. (Used [2] since they are hard to track down in the textbook, all over the place.)

$$ L^2 = \hbar^2 l(l+1) $$
$$ S^2 = \hbar^2 s(s+1) $$
$$ J^2 = \hbar^2 j(j+1) $$
$$ L_z = \hbar m $$
$$ S_z = \hbar m_s = \pm \hbar s $$

The last one is worth taking note of—$m_s$ is the spin projection quantum number which can take values of $\pm s$ in integer steps. Except spin can only be $\pm 1/2$ in this case, so we can just make the final substitution shown. 

Now we are prepared to find the exact values.

$$ (H_{00} + \lambda \vec{L}\cdot\vec{S} )|nlms> $$
$$ = (H_{00} + \lambda \frac{J^2-S^2-L^2}{2}) |nlms> $$
$$ = (e_n + \lambda\hbar^2 \frac{j(j+1)-s(s+1)-l(l+1)}{2}) |nlms> $$

Which, for a), is an exact eigenenergy of 

$$ e_n + \lambda\hbar^2 \frac{j(j+1)-s(s+1)-l(l+1)}{2} $$

Next, b). 

$$ (H_{00} + \mu L_z + \nu S_z) |nlms> $$
$$ = (e_n + \mu \hbar m \pm \nu \hbar s) |nlms> $$

Which produduces an exact eigenenergy for b) of:

$$ e_n + \mu \hbar m \pm \nu \hbar s $$

We can naturally find energies for the various states. 2S comes from n=2, l=0, m=0, s=1/2. Below is a) and b), respectively.

$$ e_2 + \lambda\hbar^2 \frac{s(s+1)-s(s+1)}{2} = e_2 $$ 
$$ e_2 \pm \nu \hbar \frac{1}{2} $$

Note how the first state has no correction at all. This is not exactly unexpected, as it was an S state, but the second one does have a correction, one that is different depending on spin. 

2P has the option for degeneracy. n=2, l=1, m=-1 or 0 or 1, and s=1/2. For a):

$$ e_2 + \lambda\hbar^2 \frac{\frac{3}{2}(\frac{3}{2}+1)-\frac{3}{4}-2}{2} $$
$$ = e_2 + \lambda\hbar^2 \frac{\frac{15}{4}-\frac{11}{4}}{2} $$
$$ = e_2 + \lambda\hbar^2 \frac{1}{2} $$

2P for b):

$$ e_2 + \mu \hbar m \pm \nu \hbar \frac{1}{2} $$

Which can acutally take on 3 values for 2P. 

$$ e_2 - \mu \hbar \pm \nu \hbar \frac{1}{2} $$
$$ e_2 \pm \nu \hbar \frac{1}{2} $$
$$ e_2 + \mu \hbar \pm \nu \hbar \frac{1}{2} $$

Note that with the spin-split, literally none of these energies are degenerate. 

Now we need to consider possible degeneracies for the perturbations. 

a)2S has 2-fold spin degeneracy. a)2P has 6-fold degeneracy for both spin and magnetic moment all together. 

b)2S has no degenreacy, for both energies are distinct. As already mentioned, b)2P has six distinct energies. 

So the b) solutions don't even need to be qeustioned, the final perturbations for them can be done with the usual theory.

$$ <nlms| \lambda \vec{L}\cdot\vec{S} |nlms> $$
$$ = \lambda <nlms| \frac{J^2-S^2-L^2}{2} |nlms> $$
$$ = \lambda\hbar^2 <nlms| \frac{j(j+1)-s(s+1)-l(l+1)}{2} |nlms> $$
$$ = \lambda\hbar^2 \frac{j(j+1)-s(s+1)-l(l+1)}{2} <nlms|nlms> $$
$$ = \lambda\hbar^2 \frac{j(j+1)-s(s+1)-l(l+1)}{2} $$

This is the general case solution. Note that it's exactly the same as a)'s exact solution. 

For 2S, the first-order perturbation correction is:

$$ \lambda\hbar^2 \frac{s(s+1)-s(s+1)}{2} = 0$$

Otherwise known as absolutely nothing, and for 2P the first-order perturbation correction is:

$$ \lambda\hbar^2 \frac{1}{2} $$

Yes this is the exact same form as a)'s corrections. As was to be expected.

Now, a) itself has a lot of degeneracies, suggesting that we might need to use degenerate perturbaiton theory. However, that turns out not to be the case, as the matrix is already diagonalized. None of our operators actually alter the ket |nlms>, they simply produce a constant eigenvalue, so in every case a <n'l'm's'|nlms> inner product is going to show up. Now, the question is, do any of these equal nonzero off hte diagonal?

Well, first of all, we know that n and l are always the same within our degeneracy, so they don't even matter, reducing our concern to <m's'|ms>. Now, we already know that the magnetic moment is orthogonal, that is, m'=m is a requirement. But is s'=s? Well, obviously, as s=$\frac{1}{2}$, however there's a little nuance here as technically we're leaving off $m_s$ in our notation here, which has a plus or minus value. The best way to check this is just to take two spiniors, one up and one down, and see what the result is.

$(0,-1) \cdot (1,0) = 0$ and $(1,0) \cdot (0,-1) = 0$

So yes, they are in fact orthogonal. Which means that none of the off-diagonal parts of the matrix are anything other than zero, meaning we can use standard perturbation theory.

Which takes the form of:

$$ <nlms| \mu L_z + \nu S_z |nlms> $$

$$ \mu \hbar m \pm \nu s $$

As before, this is just exactly the same. Rather than dragging this out, let's just state the answers that are obvoius at this point.

For 2S, the first-order perturbation correction for a) is:

$$ \pm \nu s $$

For 2P, the first-order perturbation correction for a) is

$$ e_2 + \mu \hbar m \pm \nu \hbar \frac{1}{2} $$

Which can acutally take on 3 values for 2P. 

$$ e_2 - \mu \hbar \pm \nu \hbar \frac{1}{2} $$
$$ e_2 \pm \nu \hbar \frac{1}{2} $$
$$ e_2 + \mu \hbar \pm \nu \hbar \frac{1}{2} $$

Which leads us to a final conclusion. The answers to both a) and b) for the TOTAL energy are, ultimately, the same:

$$E_{2S} = e_2 \pm \nu \hbar \frac{1}{2} $$
$$E_{2P} = e_2  + \mu \hbar m \pm \nu \hbar \frac{1}{2} + \lambda\hbar^2 \frac{1}{2} $$

Note that 2S has 2 options, and 2P has 6 (for m can be -1, 0, or 1).

Thus, we declare that it does not matter which order the two "perturbations" are done in, the end result is the same.
