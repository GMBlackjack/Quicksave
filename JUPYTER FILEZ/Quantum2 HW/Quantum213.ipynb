Now the infinite square well that goes from 0 to a has wavefunctions of the form:

$$ \psi_n = \sqrt\frac{2}{a} sin(n\pi x /a) $$

But we are in a coordinate system with the origin in the center of the well, so we have to split up into even and odd solutions. Which is to say:

$$ \psi_n = \sqrt\frac{2}{a} cos(n\pi x /a); n=odd $$
$$ \psi_n = -\sqrt\frac{2}{a} sin(n\pi x /a); n=even $$

This means we're going to have to check the expected value for combinations of even and odd functions, just in case that changes the expectation result. While m cannot equal n, that does not restrict them to not both being even or odd. Since we know what n is, 1 (ground state for square well), this leaves only m being even or odd. So we have an even-odd and an odd-odd case. This will create two not exactly trivial integrals. 

$$ \frac{2}{a} \int_{-a/2}^{a/2} cos(m\pi x/a) sin(\pi x/a) cos(\pi x/a) dx  $$
$$ \frac{2}{a} \int_{-a/2}^{a/2} -sin(m\pi x/a) sin(\pi x/a) cos(\pi x/a) dx  $$

However, we're in luck! The top equation, for m=odd, is always *odd*, which means the integral must go to zero. This means that the transition from the ground state to other odd states is "forbidden" to the first order. (Which conveninetly makes it so we don't have to worry about the m=n case at all, as that term vanishes!)

This leaves just the one integral:

$$ \frac{2}{a} \int_{-a/2}^{a/2} -sin(m\pi x/a) sin(\pi x/a) cos(\pi x/a) dx  $$

And thus we return to our older equations and work back to the start:

$$V_{21} =  -\frac{V_0\tau}{\pi(t^2+\tau^2)} $$

$$V_{21}(\omega_{21}) = -\int^\infty_{-\infty} \frac{1}{2\pi} \frac{V_0\tau}{\pi(t^2+\tau^2)} e^{i\omega t} dt$$
$$ = -\frac{V_0\tau}{2\pi^2}\int^\infty_{-\infty} \frac{1}{(t^2+\tau^2)} e^{i\omega t} dt$$

Which we then integrate, wait, integrating this is quite the pain, let's refer to Wolfram's Fourier Transform [1], for that's what this is. 

The constants out front are a bit different, so we need to be careful when substituting the solution in. However, once we do this, we get:

$$V_{21}(\omega_{21}) = -\frac{V_0\tau}{2\pi}\frac{e^{-\tau|\omega|}}{\tau} = -\frac{V_0}{2\pi}e^{-\tau|\omega|} $$

We had been mildly concerned about the negative sign, but the next step has us take the square magnitude, so all is good. 

$$P_{21} = \frac{4\pi^2}{\hbar^2} \frac{V_0^2}{4\pi^2}e^{-2\tau|\omega|} $$

Thus, the probability that it is the nth eigenstate is **zero**, unless n=1 or n=2. In which case, the probability for n=2 is 

$$P_{21} = \frac{V_0^2}{\hbar^2}e^{-2\tau|\omega|} $$

And it remaining in the ground state is:

$$P_{11} = 1 - \frac{V_0^2}{\hbar^2}e^{-2\tau|\omega|} $$

All else is "forbidden" to first order.
