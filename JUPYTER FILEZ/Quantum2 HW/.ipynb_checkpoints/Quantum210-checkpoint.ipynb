## Problem 1



First, let's factorize the polynomial in the denominator. Clearly, we'll need to use imaginary numbers so rather than trying to guess it by pattern recognition let's just feed it into a root finder [1] producing -2+i and -2-i as our roots, meaning the integral can now be written as

$$ \int^\infty_{-\infty} \frac{sinx}{(x+2+i)(x+2-i)} dx $$

The standard Contour for Contour Integration is a semicircle that goes above the physical (x) axis into the imaginary plane. The loop created by this shape contains the point where -2+i resides, but not -2-i, so we only have to worry about one of the points: -2+i. As it turns out, though, this is not strictly true. We shall see why if we split the sine function up:

$$ \int^\infty_{-\infty} \frac{e^{ix}-e^{-ix}}{2i(x+2+i)(x+2-i)} dx $$

We will later have to integrate over the complex plane, and with the usual $z=Re^{i\theta}$ either one exponential or the other will blow up as R grows without bound. However, there is a solution to this: make two integrals. 

$$ \int^\infty_{-\infty} \frac{e^{ix}}{2i(x+2+i)(x+2-i)} dx - \int^\infty_{-\infty} \frac{e^{-ix}}{2i(x+2+i)(x+2-i)} dx$$

One of these we will use an integral over a semicircle in the +i part of the complex plane, and in the other we will use in the -i part of the plane. Each one will have a single residue. 

The best part is that each individual semicircle will go to zero: x will be replaced with z=$Re^{i\theta}$, but the bounds of integration will change. This will make one produce positive-imaginary results, and the other will get negative-imaginary results. We will put the negative imaginary results in the left integral in order to get something of the form $e^{-iR}$ which, as R approaches infinity, goes to zero everywhere. The positive-imaginary goes to the other side. Thus, we can now say that the semicircular parts of our contours will go to zero, meaning the entire contour's integral will equal the real line's integral. 

Then we will add the two together to get the total. 

The residue theorem gives us

$$\oint_C f(z)dz=2\pi ia_{-1}$$

for single poles, which is precisely what we have. To calculate the residue, we use

$$a_{-1} = lim_{z\rightarrow z_0}(z-z_0)f(z)$$

Which is to say, for the first residue, divide out the zero then insert -2+i. 

$$a_{-1a} = lim_{z\rightarrow -2+i}(z+2-i) \frac{e^{iz}}{2i(z+2+i)(z+2-i)} $$
$$ = lim_{z\rightarrow -2+i} \frac{e^{iz}}{2i(z+2+i)} $$
$$ = \frac{e^{-2i-1}}{-4} $$

For the second case, -2-i. 

$$a_{-1a} = lim_{z\rightarrow -2-i}(z+2+i) \frac{e^{-iz}}{2i(z+2+i)(z+2-i)} $$
$$ = lim_{z\rightarrow -2-i} \frac{e^{-iz}}{2i(z+2-i)} $$
$$ = \frac{e^{2i-1}}{4} $$

A little ugly since we're passing whole numbers to exponential functions, not multiples of $\pi$, so we shall avoid evaluating as long as we can. 

Regardless, we can now state our contour integrals' results.

$$ \oint_C f(z)dz = 2 \pi i \frac{e^{-2i-1}}{-4}  $$
$$ - \oint_C f(z)dz = 2 \pi i \frac{e^{2i-1}}{4}  $$

NOTE THE MINUS SIGN: one of the contours needs to be performed BACKWARD since we need to travel along the real number line in the *same direction*.

Subtraction provides:

$$ 2 \pi \left( \frac{ie^{-2i-1}}{-4}-\frac{ie^{-2i-1}}{-4} \right) $$
$$=  \pi \left( \frac{e^{-2i-1}-e^{-2i-1}}{2i} \right) $$
$$ =  \frac{\pi} {e} \left( \frac{e^{-2i}-e^{-2i}}{2i} \right) $$
$$ = \frac{\pi} {e} sin(-2) $$

## Problem 2



The Born Approximation is:

$$ f_{\vec{k}}(\hat{r}) \approx -\frac{m}{2\pi\hbar^2} \int e^{i(\vec{k}-\vec{k}') \cdot \vec{r}'} V(r) d^3r'  = f_{\vec{k}}^{BORN} $$
$$ = \frac{m}{2\pi\hbar^2} \int e^{i(\vec{k}-\vec{k}') \cdot \vec{r}'} V_0e^{-r'/a} d^3r'  = f_{\vec{k}}^{BORN} $$

This is somewhat similar (but not identical) to the Yukawa potential which is solved for on page 386.

Since V is a spherically symmetric potential, we can use equation 6.74 for the born approximation:

$$ = \frac{2m}{q\hbar^2} \int_0^\infty r V_0e^{-r/a} sin(qr) dr $$

Where $q = |\vec{k}-\vec{k}'| = 2ksin(\theta/2)$. This step is effectively the same as defining the z axis of integration in the r direction.

``ASIDE BEGIN``
It's not necessary but in order to facilitate understanding, let's actually work this out. First of all, what is q? 

q is actually a magnitude of the vector $\vec{q} = (\vec{k}-\vec{k'})$. The angle $\theta$ is the angle between k and k', so we are able to construct an equation relating the magnitude q to the angle: specifically $2ksin(\theta/2)$. Just to know where that comes from. 

Now we take the Born approximation on its own.

$$ f_{\vec{k}}^{BORN} = -\frac{m}{2\pi\hbar^2} \int e^{i(\vec{k}-\vec{k}') \cdot \vec{r}'} V(r) d^3r'  $$

We convert into spherical, defining the z-direction as $k\hat{r}-\vec{k}$ as suggested, which physically means the z-direction actually points in the direction of $\vec{q}$ now. which conveniently makes the angular components of the $\vec{q}$ zero since its arrow is pointing in the z-direction, where $\theta=0$ and $\phi$ is irrelevant. This removes the vector dependence and gives us the ability to integrate in spherical terms.

$$ f_{\vec{k}}^{BORN} = -\frac{m}{2\pi\hbar^2} \int e^{iqr} V(r) r^2 sin\theta drd\theta d\phi  $$

Since we're integrating we do need to write q out. Also $\phi$ is easy.

$$ f_{\vec{k}}^{BORN} = -\frac{m}{\hbar^2} \int e^{2iksin(\theta/2)r} V(r) r^2 sin\theta drd\theta  $$

``ASIDE END``

Regardless, we now evaluate the integral using [3]

$$ = \frac{2mV_0}{q\hbar^2} \left[ -\frac{ae^{-r/a}(((a^2q^2+1)r-a^3q^2+a)sin(qr)+aq(a^2q^2r+r+2a)cos(qr))}{(a^2q^2+1)^2} \right]_0^\infty$$

$$ = \frac{2mV_0}{q\hbar^2} \left[ -0 + \frac{a(0+aq(0+0+2a))}{(a^2q^2+1)^2} \right]$$
$$ = \frac{2mV_0}{q\hbar^2} \left[ \frac{2a^3q}{(a^2q^2+1)^2} \right]$$
$$ = \frac{4mV_0a^3}{\hbar^2} \left[ \frac{1}{(a^2q^2+1)^2} \right]$$

Square this (as it has no imaginary component) and replace q with its definition in terms of the angle.

$$ D(\theta) = \frac{16m^2V_0^2a^6}{\hbar^4} \left[ \frac{1}{(a^2q^2+1)^4} \right]$$
$$  = \frac{16m^2V_0^2a^6}{\hbar^4} \left[ \frac{1}{(4k^2a^2sin^2(\theta/2)+1)^4} \right]$$

If we were willing to use some approximations, we could probably pull out the k in the denominator, but so far as we can tell we want an exact answer, so this is it. 

## Problem 3



Start with the general solution to the radial wave equation for scattering. 

$$ R_{kl} = cos\delta_l(k)j_l(kr) - sin\delta_l(k)n_l(kr) $$

We want an expression for $d_l$. However, to get anywhere with this, we need $R_{kl}$ for the inner portion, which is the solution to the radial equation for a $1/r^2$ potential. We were certain we had solved the radial equation for $1/r^2$ before, but it seems not, so we shall do it. So we return to our friend the modified radial wave equation:

$$ \left[  -\frac{\hbar^2}{2m} \frac{d^2}{dr^2} + \frac{l(l+1)\hbar^2}{2mr^2} + V(r) \right]u = Eu $$

$$ \Rightarrow \left[  -\frac{\hbar^2}{2m} \frac{d^2}{dr^2} + \frac{l(l+1)\hbar^2}{2mr^2} + \frac{C}{r^2} \right]u = Eu $$

$$ \Rightarrow \left[  -\frac{\hbar^2}{2m} \frac{d^2}{dr^2} + (\frac{l(l+1)\hbar^2}{2m}+C)\frac{1}{r^2} \right]u = Eu $$

$$ \Rightarrow \left[  -\frac{\hbar^2}{2m} \frac{d^2}{dr^2} + (\frac{l(l+1)\hbar^2}{2m}+C)\frac{1}{r^2} - E \right]u = 0  $$

$$ \Rightarrow \left[  - \frac{d^2}{dr^2} + (l(l+1)\hbar^2+2mC)\frac{1}{\hbar^2r^2} - \frac{2mE}{\hbar^2} \right]u = 0  $$

$$ \Rightarrow \left[ \frac{d^2}{dr^2} + \frac{2mE}{\hbar^2} - (l(l+1)\hbar^2+2mC)\frac{1}{\hbar^2r^2} \right]u = 0  $$

Of course the difficulty arises in actually solving this for u. (and then obtaining R from u=rR.)

Remember the usual boundary condition: at r=0 the result must be zero, usually accomplished with u(0)=0. 

Simplify the above as with the hydrogen atom, condensing energy into $k^2 = 2mE/\hbar^2$. Adding a length scale wouldn't do much as a lot of the constants are different between terms. 

$$ \Rightarrow \left[ \frac{d^2}{dr^2} + k^2 - (l(l+1)\hbar^2+2mC)\frac{1}{\hbar^2r^2} \right]u = 0  $$

This is remarkably close to the modified radial equation for constant potential.

$$ \left[ \frac{d^2}{dr^2} + k^2 - \frac{l(l+1)}{r^2}  \right]u=0 $$

Which we know has a solution R = Aj+Bn or u = Aj/r + Bn/r, where j and n are the spherical bessel functions. The only issue is that j and n are defined in terms of $j_l$, that is, l matters, and we have drastically altered how the l behaves in the function. However, if we could get it into the same form, we could figure out what the subscript SHOULD be. Multiplying it all out, and having the "old" l values represented by L, we have...

$$ l^2 + l + \frac{2mC}{\hbar^2} = L^2 + L $$

The idea is to solve for L. Curiously this just ends up being an application of the quadratic formula. 

$$ L = \frac{-1 \pm \sqrt{1-4(-l^2-l-\frac{2mC}{\hbar^2})}}{2} $$

The plus or minus provides some annoyance, as while the resulting subscript being negative isn't a problem as l could always do that anyway, it's *unbalanced* so they aren't equivalent. Perhaps it will become evident later which sign to use...

Anyway, the standard solution for the spherical Bessel functions is $Aj_l + Bn_l$. Since our potential covers r=0 we can set B=0 and the solution is $Aj_l$. For *our* solution, we replace l with L. And do not write it out because that would be the ugliest subscript in the universe.

$$ R = Aj_L $$

This isn't normalized but let's not worry about that right now, A is just some constant.

Now we relate the outside R to it's large-r behavior in the slides, we have...

$$ R = \frac{sin(kr-l\pi/2+\delta_l)}{kr} $$

So now we take the logarithmic derivative of both Rs and set them equal to each other at r=a. Reminder: Logarithmic Derivative of f is f'/f. For the "inner" part:

$$ kj_L'(kr)/j_L(kr) $$

And for the "outer" part:

$$ \frac{-sin(kr-l\pi/2+\delta_l)+krcos(kr-l\pi/2+\delta_l)}{kr^2} \frac{kr}{sin(kr-l\pi/2+\delta_l)} $$

$$  = \frac{-sin(kr-l\pi/2+\delta_l)+krcos(kr-l\pi/2+\delta_l)}{rsin(kr-l\pi/2+\delta_l)} $$
$$  = \frac{-1}{r}+\frac{k}{tan(kr-l\pi/2+\delta_l)} $$

We now proceed to set the two to be equal to each other and solve for delta. At the arbitrary boundary a. 

$$ \frac{kj_L'(ka)}{j_L(ka)}  = \frac{-1}{a}+\frac{k}{tan(ka-l\pi/2+\delta_l)} $$
$$ \frac{kj_L'(ka)}{j_L(ka)} + \frac{1}{a} = \frac{k}{tan(ka-l\pi/2+\delta_l)} $$
$$ \frac{1}{\frac{kj_L'(ka)}{j_L(ka)} + \frac{1}{a}} = \frac{tan(ka-l\pi/2+\delta_l)}{k} $$
$$ \frac{k}{\frac{kj_L'(ka)}{j_L(ka)} + \frac{1}{a}} = tan(kr-l\pi/2+\delta_l) $$
$$ atan\left(\frac{k}{\frac{kj_L'(ka)}{j_L(ka)} + \frac{1}{a}}\right) = kr-l\pi/2+\delta_l $$
$$ atan\left(\frac{k}{\frac{kj_L'(ka)}{j_L(ka)} + \frac{1}{a}}\right) - kr + l\pi/2 = \delta_l $$

Now we ask what sign L has inside it's definition. 

From figure 6.9 on pg 399 we see that the "phase adjustment" or "displacement" from the center is equal to $-\delta_0/k$, letting us know that the actual value of $\delta_0$ better be positive or else it would be displaced *into* the center, which we know isn't right. For l=0, we end up with atan(function)-kr, and we want the result to be positive, which is only accomplished if L itself is positive. (Furthermore the quantum number l itself is always positive). Therefore, we submit as our answer:

$$ atan\left(\frac{k}{\frac{kj_L'(ka)}{j_L(ka)} + \frac{1}{a}}\right) - kr + l\pi/2 = \delta_l $$

with

$$ L = \frac{-1 + \sqrt{1+4(l^2+l+\frac{2mC}{\hbar^2})}}{2} $$

Now we examine the cross section via equation 6.127

$$ \sigma =  \frac{4\pi}{k^2} \sum_{l=0}^\infty (2l+1)sin^2(\delta_l) $$

As l increases going towards infinity, it dominates the form of $\delta_l$ Thus the sum becomes

$$ \sigma =  \frac{4\pi}{k^2} \sum_{l=0}^\infty (2l+1)sin^2(\frac{l\pi}{2}) $$

l can only take integer values, and for every odd value it will make the $sin^2$ function equal 1, and for every even value it will equal 0. It will never converge, effectively adding 0+1+0+1+0... forever as l gets larger and larger. Thus, the series is **divergent**.

Logically, since $C/r^2$ is a repulsive potential, there will be many fewer particles scattering at the small angles and more at the large ones, thus it would have to be the large angles that provide the divergence, as they have to contribute more than the small angles.
