A reminder that "totally forbidden" does not actually mean "totally forbidden" it just means its not possible in the approximation we are using. Such transitions will occur in reality.

"Forbidden" means that $V_{mn} = 0$, that is the transition from the n-state to the m-state does not occur. "Totally forbidden" is a stronger condition than this, specifically refering to an expectation value being zero without a series expansion approximation. In our specific case, this means the electric dipole approximation is NOT used and we end up with:

$$ \langle \psi_{m00} | e^{i\omega_{mn}\hat{n}\cdot\vec{r}/c} \vec{p} \cdot \hat{e} | \psi_{n00} \rangle  $$

And this is what we need to show always equals zero. We are not allowed to approximate the exponential. 

We already know that, by pairity, $ \langle \psi_{m00} | \vec{p} \cdot \hat{e} | \psi_{n00} \rangle = 0 $. The question is, what could the exponential term possibly change in regards to that? 

It has become evident (over email) that using mathematical procedures to prove this relation is out of the scope of this course, and we instead need to use logical arguments. So what do we know about the behavior of the imaginary exponential?

One thing we note is that we can rewrite the exponential $e^{i\omega_{mn}\hat{n}\cdot\vec{r}/c}$ as $e^{i\omega_{mn}r cos\theta/c}$ because $\hat{n}$ is a unit vector, and we can set it along the z-axis without loss of generality to make the angle easier to deal with. 

Converting to integral form has been determined to be a waste of time so, instead, we rely on the commutator $[H_0,\vec{r}] = \frac{\hbar}{im}\vec{p}$ to write...

$$ = \langle \psi_{m00} | e^{i\omega_{mn}\hat{n}\cdot\vec{r}/c} \frac{im}{\hbar} \left( H_0\vec{r}-\vec{r}H_0 \right) \cdot \hat{e} | \psi_{n00} \rangle  $$
$$ = \frac{im}{\hbar} \langle \psi_{m00} | e^{i\omega_{mn}\hat{n}\cdot\vec{r}/c} \left( H_0\vec{r}-\vec{r}H_0 \right) \cdot \hat{e} | \psi_{n00} \rangle  $$

Introduce the complicated one.

$$ = \sum_a \frac{im}{\hbar} \langle \psi_{m00} | e^{i\omega_{mn}\hat{n}\cdot\vec{r}/c} |\psi_{a00} \rangle\langle \psi_{a00}| \left( H_0\vec{r}-\vec{r}H_0 \right) \cdot \hat{e} | \psi_{n00} \rangle  $$
$$ = \sum_a \frac{im}{\hbar} \langle \psi_{m00} | e^{i\omega_{mn}\hat{n}\cdot\vec{r}/c} |\psi_{a00} \rangle (E_a-E_n)\langle \psi_{a00}| \vec{r} \cdot \hat{e} | \psi_{n00} \rangle  $$

Now, we already *know* the result for the second term, $<\vec{r}\cdot\hat{e}>$, is zero for s-states by the pairity argumetn since $\Delta l = 0$. (Everything is an s-state, all l terms equal zero, thus the pairity argument holds).

With it equaling zero, the exponential term *doesn't matter* as suddenly the entire sum is zero, being multiplied by zero, and thus everything is Totally Forbidden.

The electric quadrupole term is

$$ M^{eq}_{mn} = -\frac{m\omega^2_{mn}}{2c} \langle \psi_m | \hat{e} \cdot \vec{r} \hat{n} \cdot \vec{r} | \psi_n \rangle $$

In the end this should be able to scale directly to the dipole term by some constant ratio. The question is how exactly we evaluate it. 

We consider ourselves to be initially in the ground state, that is $|n_xn_yn_z\rangle = |000\rangle$. To make this easier let's also set the e-direction to the z-direction without loss of generality. This doesn't change much, but it does give us:

$$ M^{eq}_{mn} = -\frac{m\omega^2_{mn}}{2c} \langle n_xn_yn_z | \hat{z} \cdot \vec{r} \hat{n} \cdot \vec{r} | 000 \rangle $$

The trick is to, somehow, evaluate these matrix elements. We can actually evaluate the left dot product...

$$ M^{eq}_{mn} = -\frac{m\omega^2_{mn}}{2c} \langle n_xn_yn_z | z (\hat{n} \cdot \vec{r}) | 000 \rangle $$

We remember that the e and n directions are pe4rpendicular, so we choose to set the n direction to x. This allows us to reduce the seocnd dot product to:

$$ M^{eq}_{mn} = -\frac{m\omega^2_{mn}}{2c} \langle n_xn_yn_z | z x | 000 \rangle $$

Note that we can actually split this up: each wavefunction is actually a combination of the x, y, and z parts. 

$$ = -\frac{m\omega^2_{mn}}{2c} \langle n_x | \langle n_y | \langle n_z | zx |0\rangle |0\rangle |0\rangle $$

Each operator will interact with the wavefunctions surrounding them based on their orietnation relative to the harmonic oscillator. That is, for the x-direction, only the x-operator will interact, y and z will "pass through" so to speak. From griffiths, we know we can write out the x-operator in terms of just raising and lowering operators:

$$ x = \sqrt{\frac{\hbar}{2m\omega}}(a^+ + a^-) $$

Where the operators themselves act on their in-line functions as

$$ a^+|n> = \sqrt{n+1}|n+1> ; a^-|n> = \sqrt{n}|n-1>$$

The nice thing here is that the lowering operator is always going to act on the |0> ket, and we know that lowering on the |0> ket produces a 0 term as it is at the "bottom of the ladder" so we end up only having to deal with one term--the raising operator. The z-operator will interact with the z-functions, the x-operator will interact with the x-functions, and everything else does nothing. So we end up with...

$$ = -\frac{m\omega^2_{mn}}{2c}  \sqrt{\frac{\hbar}{2m\omega}}\langle n_x | 1 \rangle \langle n_y | 0 \rangle  \sqrt{\frac{\hbar}{2m\omega}} \langle n_z | 1 \rangle$$

$$ = -\frac{m\omega^2_{mn}\hbar}{4m\omega c}  \langle n_x | 1 \rangle \langle n_y | 0 \rangle \langle n_z | 1 \rangle $$
$$ = -\frac{\omega^2_{mn}\hbar}{4\omega c}  \delta_{n_x1}\delta_{n_y0}\delta_{n_z1} $$

$$A sin(kr_1) e^{-Qr_2} r_1^2 \rightarrow -\frac{2A(3Q^2k-k^3)}{(3Qk^2-Q^2)^2+(3Q^2k-k^2)^2} $$
$$A sin(kr_1) e^{-Qr_2} r_1^3 \rightarrow -\frac{6A(4Q^2k-4Qk^3)}{(4kQ^3-4k^3Q)^2+(k^4+Q^4-6Q^2k^2)^2} $$
$$A sin(kr_1) e^{-Qr_2} r_1^4 \rightarrow -\frac{24A(k^5-10Q^2k^3-5Q^4k)}{(Q^5-5Qk^4+10Q^3k^2)^2+(k^5-10Q^2k^3+5Q^4k)^2} $$


Let Q=4/a and let q=1/a

This means the actual, full, complete answer is:

$$ \frac{6(4Q^2k-4Qk^3)}{(3(4kQ^3-4k^3Q)^2+(k^4+Q^4-6Q^2k^2)^2)} -\frac{2(a+1)(3Q^2k-k^3)}{(3(3Qk^2-Q^2)^2+(3Q^2k-k^2)^2)} + \frac{akQ}{(3(2kQ)^2+(Q^2-k^2)^2)} + ... + \frac{6kqa^2}{(3^4(2kq)^2+(q^2-k^2)^2)} $$

Notably we did not bother to write out every single term since simply writing out these four terms took an absurd amount of time. We are under no illusions that this is the correct answer.
