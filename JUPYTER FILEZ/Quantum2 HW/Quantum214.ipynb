$$ = -\frac{8 \pi^2 e^2}{ki} \int_0^\infty  \left[(e^{ikr_1}-e^{-ikr_1})\left( \frac{r_1^3-r_1^2(a+1)+r_1a}{3}+\frac{3ar_1^2+r_1(-2a^2-2a)+a^2}{3^2}+\frac{6r_1a^2-2a^3-2a^2}{3^3}+\frac{6a}{3^4} - \frac{r_1^4(1/a)-r_1^3(1+1/a)+r_1^2}{3}-\frac{3r_1^3+r_1^2(-2a-2)+ar^1}{3^2}-\frac{6r_1^2a-2r_1a^2-2r_1a}{3^3}-\frac{6r_1}{3^4} \right)e^{-4r_1/a} + (e^{ikr_1}-e^{-ikr_1})\left( \frac{2a^3}{3^3} - \frac{6a^3}{3^4} - \frac{2r_1a^2}{3^3} + \frac{6r_1a^2}{3^4} \right)e^{-r_1/a}\right] dr_1 $$

But this now makes everything of the form $Ar_1^n e^{-Qr_1}$ where Q and A are variable coefficients and n is an integer. Integrating each of these things has a very specific result at $r_1$ = 0. 

Here is every case at zero: 

$$A  e^{-Qr_1} r_1^0 \rightarrow -\frac{A}{Q} $$
$$A  e^{-Qr_2} r_1^1 \rightarrow -\frac{A}{Q^2} $$
$$A  e^{-Qr_2} r_1^2 \rightarrow -\frac{2A}{Q^3} $$
$$A  e^{-Qr_2} r_1^3 \rightarrow -\frac{6A}{Q^4} $$
$$A  e^{-Qr_2} r_1^4 \rightarrow -\frac{24A}{Q^5} $$

Every single term will evaluate in this manner. 

Now, the evaluation will contain a lot of terms. All of them will have their signs flipped since the 0 is the bottom of the itnegral. Every A will be unique. However, there are a set number of Q values:

(-ik+4/a),(ik+4/a),(-ik+1/a),(ik+1/a). 

Which we note are just complex numbers. And so the result is...

$$ -\frac{8\pi^2e^2}{ki} \left[ \frac{6}{3(-ik+4/a)^4}+\frac{2a+2}{3(-ik+4/a)^3}+\frac{a}{3(-ik+4/a)^2}+\frac{6a}{3^2(-ik+4/a)^3}+\frac{(-2a^2-2a)}{3^2(-ik+4/a)^2}+\frac{a^2}{3^2(-ik+4/a)}+\frac{12a^2}{3^3(-ik+4/a)^2}+\frac{-2a^3-2a^2}{3^3(-ik+4/a)}+\frac{6a}{3^4(-ik+4/a)}-\frac{24(1/a)}{3(-ik+4/a)^5}-\frac{6(1+1/a)}{3(-ik+4/a)^4}-\frac{2}{3(-ik+4/a)^3}-\frac{18}{3^2(-ik+4/a)^4}-\frac{-4a-4}{3^2(-ik+4/a)^3}-\frac{a}{3^2(-ik+4/a)^2}-\frac{12a}{3^3(-ik+4/a)^3}-\frac{-2a^2-2a}{3^3(-ik+4/a)^2}-\frac{6}{3^4(-ik+4/a)^2} + \frac{6}{3(ik+4/a)^4}+\frac{2a+2}{3(ik+4/a)^3}+\frac{a}{3(ik+4/a)^2}+\frac{6a}{3^2(ik+4/a)^3}+\frac{(-2a^2-2a)}{3^2(ik+4/a)^2}+\frac{a^2}{3^2(ik+4/a)}+\frac{12a^2}{3^3(ik+4/a)^2}+\frac{-2a^3-2a^2}{3^3(ik+4/a)}+\frac{6a}{3^4(ik+4/a)}-\frac{24(1/a)}{3(ik+4/a)^5}-\frac{6(1+1/a)}{3(ik+4/a)^4}-\frac{2}{3(ik+4/a)^3}-\frac{18}{3^2(ik+4/a)^4}-\frac{-4a-4}{3^2(ik+4/a)^3}-\frac{a}{3^2(ik+4/a)^2}-\frac{12a}{3^3(ik+4/a)^3}-\frac{-2a^2-2a}{3^3(ik+4/a)^2}-\frac{6}{3^4(ik+4/a)^2} + \frac{2a^3}{3^3(-ik+1/a)}-\frac{6a^3}{3^4(-ik+1/a)}-\frac{2a^2}{3^3(-ik+1/a)^2}+\frac{6a^2}{3^4(-ik+1/a)^2} + \frac{2a^3}{3^3(ik+1/a)}-\frac{6a^3}{3^4(ik+1/a)}-\frac{2a^2}{3^3(ik+1/a)^2}+\frac{6a^2}{3^4(ik+1/a)^2} \right] $$

Keep in mind that the final answer is of the form:

$$ W = \frac{\sqrt{\pi}m^{3/2}}{\hbar^4} \sqrt{-E_{He}} Z^{9}a^{-9} | IntegralGoesHere |^2 $$

We can pull out the leading constant and at least try to simplify the outside. EVery term has a 1/3 in it so we can pull that out as well. 

$$ W = \frac{\sqrt{\pi}m^{3/2}}{\hbar^4} \sqrt{-E_{He}} 2^{9}a^{-9} \frac{2^6\pi^4e^4}{k^23^2} | IntegralInterior |^2 $$
$$ W = \frac{2^{16}\pi^{9/2}m^{3/2}e^4\sqrt{-E_{He}}}{3^2k^2\hbar^4a^9} | IntegralInterior |^2 $$


The integral interior is now

$$ \frac{6}{(-ik+4/a)^4}+\frac{2a+2}{(-ik+4/a)^3}+\frac{a}{(-ik+4/a)^2}+\frac{6a}{3(-ik+4/a)^3}+\frac{(-2a^2-2a)}{3(-ik+4/a)^2}+\frac{a^2}{3(-ik+4/a)}+\frac{12a^2}{3^2(-ik+4/a)^2}+\frac{-2a^3-2a^2}{3^2(-ik+4/a)}+\frac{6a}{3^3(-ik+4/a)}-\frac{24(1/a)}{(-ik+4/a)^5}-\frac{6(1+1/a)}{(-ik+4/a)^4}-\frac{2}{(-ik+4/a)^3}-\frac{18}{3(-ik+4/a)^4}-\frac{-4a-4}{3(-ik+4/a)^3}-\frac{a}{3(-ik+4/a)^2}-\frac{12a}{3^2(-ik+4/a)^3}-\frac{-2a^2-2a}{3^2(-ik+4/a)^2}-\frac{6}{3^3(-ik+4/a)^2} + \frac{6}{(ik+4/a)^4}+\frac{2a+2}{(ik+4/a)^3}+\frac{a}{(ik+4/a)^2}+\frac{6a}{3(ik+4/a)^3}+\frac{(-2a^2-2a)}{3(ik+4/a)^2}+\frac{a^2}{3(ik+4/a)}+\frac{12a^2}{3^2(ik+4/a)^2}+\frac{-2a^3-2a^2}{3^2(ik+4/a)}+\frac{6a}{3^3(ik+4/a)}-\frac{24(1/a)}{(ik+4/a)^5}-\frac{6(1+1/a)}{(ik+4/a)^4}-\frac{2}{(ik+4/a)^3}-\frac{18}{3(ik+4/a)^4}-\frac{-4a-4}{3(ik+4/a)^3}-\frac{a}{3(ik+4/a)^2}-\frac{12a}{3^2(ik+4/a)^3}-\frac{-2a^2-2a}{3^2(ik+4/a)^2}-\frac{6}{3^3(ik+4/a)^2} + \frac{2a^3}{3^2(-ik+1/a)}-\frac{6a^3}{3^3(-ik+1/a)}-\frac{2a^2}{3^2(-ik+1/a)^2}+\frac{6a^2}{3^3(-ik+1/a)^2} + \frac{2a^3}{3^2(ik+1/a)}-\frac{6a^3}{3^3(ik+1/a)}-\frac{2a^2}{3^2(ik+1/a)^2}+\frac{6a^2}{3^3(ik+1/a)^2} $$

Some of these terms can be simplified.

$$ = \frac{6}{(-ik+4/a)^4}+\frac{2a+2}{(-ik+4/a)^3}+\frac{a}{(-ik+4/a)^2}+\frac{2a}{(-ik+4/a)^3}+\frac{(-2a^2-2a)}{3(-ik+4/a)^2}+\frac{a^2}{3(-ik+4/a)}+\frac{4a^2}{3(-ik+4/a)^2}+\frac{-2a^3-2a^2}{3^2(-ik+4/a)}+\frac{2a}{3^2(-ik+4/a)}-\frac{24(1/a)}{(-ik+4/a)^5}-\frac{6(1+1/a)}{(-ik+4/a)^4}-\frac{2}{(-ik+4/a)^3}-\frac{6}{(-ik+4/a)^4}-\frac{-4a-4}{3(-ik+4/a)^3}-\frac{a}{3(-ik+4/a)^2}-\frac{4a}{3(-ik+4/a)^3}-\frac{-2a^2-2a}{3^2(-ik+4/a)^2}-\frac{2}{3^2(-ik+4/a)^2} + \frac{6}{(ik+4/a)^4}+\frac{2a+2}{(ik+4/a)^3}+\frac{a}{(ik+4/a)^2}+\frac{2a}{(ik+4/a)^3}+\frac{(-2a^2-2a)}{3(ik+4/a)^2}+\frac{a^2}{3(ik+4/a)}+\frac{4a^2}{3(ik+4/a)^2}+\frac{-2a^3-2a^2}{3^2(ik+4/a)}+\frac{2a}{3^2(ik+4/a)}-\frac{24(1/a)}{(ik+4/a)^5}-\frac{6(1+1/a)}{(ik+4/a)^4}-\frac{2}{(ik+4/a)^3}-\frac{6}{(ik+4/a)^4}-\frac{-4a-4}{3(ik+4/a)^3}-\frac{a}{3(ik+4/a)^2}-\frac{4a}{3(ik+4/a)^3}-\frac{-2a^2-2a}{3^2(ik+4/a)^2}-\frac{2}{3^2(ik+4/a)^2} + \frac{2a^3}{3^2(-ik+1/a)}-\frac{2a^3}{3^2(-ik+1/a)}-\frac{2a^2}{3^2(-ik+1/a)^2}+\frac{2a^2}{3^2(-ik+1/a)^2} + \frac{2a^3}{3^2(ik+1/a)}-\frac{2a^3}{3^2(ik+1/a)}-\frac{2a^2}{3^2(ik+1/a)^2}+\frac{2a^2}{3^2(ik+1/a)^2} $$

Wow, now a rather large number of terms just up and cancel. All of the secondary exponentials with 1/a are gone. Other terms dissappear (which they would have dissapeared before but now we're actually checking for them.)

$$ = \frac{4a}{(-ik+4/a)^3}+\frac{a}{(-ik+4/a)^2}+\frac{(2a^2-3a)}{3(-ik+4/a)^2}+\frac{a^2}{3(-ik+4/a)}+\frac{-2a^3-2a^2+2a}{3^2(-ik+4/a)}-\frac{24(1/a)}{(-ik+4/a)^5}-\frac{6(1+1/a)}{(-ik+4/a)^4}-\frac{-4}{3(-ik+4/a)^3}-\frac{-2a^2-2a-2}{3^2(-ik+4/a)^2}+\frac{4a}{(ik+4/a)^3}+\frac{a}{(ik+4/a)^2}+\frac{(2a^2-3a)}{3(ik+4/a)^2}+\frac{a^2}{3(ik+4/a)}+\frac{-2a^3-2a^2+2a}{3^2(ik+4/a)}-\frac{24(1/a)}{(ik+4/a)^5}-\frac{6(1+1/a)}{(ik+4/a)^4}-\frac{-4}{3(ik+4/a)^3}-\frac{-2a^2-2a-2}{3^2(ik+4/a)^2}$$

Before we continue, we note that the value of k is known. We found it in the process of finding the density of states, we just didn't realize it. The interior of the delta function was:

$$ k^2 + \frac{2m}{\hbar^2}(E_{He}-E_i) $$

So the only k values that are possible are:

$$ k = \pm \sqrt{\frac{2m}{\hbar^2}(E_i-E_{He})} $$
$$ k = \pm \sqrt{\frac{-mE_{He}}{\hbar^2}} $$

We can relate this to the bohr radius via $1/E = 2ma^2/\hbar^2$ which gives us $a = \sqrt{\frac{\hbar^2}{2mE}}$ which allows us to say $k = \frac{\sqrt{2}}{a}$. There is some confusion on the sign, but it turns out to be irrelevant because in our purpose every k also has an identical -k elsewhere. This allows us to make the denominators easier to deal with, and move an a into the numerator. Many more terms will be able to combine. (-ik+4/a) becomes (-i√2+4)/a and the other becomes (i√2+4)/a

$$ = \frac{4a^4}{(-i√2+4)^3}+\frac{a^3}{(-i√2+4)^2}+\frac{(2a^4-3a^3)}{3(-i√2+4)^2}+\frac{a^3}{3(-i√2+4)}+\frac{-2a^4-2a^3+2a^2}{3^2(-i√2+4)}-\frac{24}{(-i√2+4)^5}-\frac{6(a^4+a^3)}{(-i√2+4)^4}-\frac{-4a^3}{3(-i√2+4)^3}-\frac{-2a^4-2a^3-2a^2}{3^2(-i√2+4)^2}+\frac{4a^4}{(i√2+4)^3}+\frac{a^3}{(i√2+4)^2}+\frac{(2a^4-3a^3)}{3(i√2+4)^2}+\frac{a^3}{3(i√2+4)}+\frac{-2a^4-2a^3+2a^2}{3^2(i√2+4)}-\frac{24}{(i√2+4)^5}-\frac{6(a^4+a^3)}{(i√2+4)^4}-\frac{-4a^3}{3(i√2+4)^3}-\frac{-2a^4-2a^3-2a^2}{3^2(i√2+4)^2}$$

And then we'd plug it into here:

$$ W = \frac{2^{16}\pi^{9/2}m^{3/2}e^4\sqrt{-E_{He}}}{3^2k^2\hbar^4a^9} | IntegralInterior |^2 $$

The math that needs to be done is evident, just tedious. Clearly there is either another way to simplify this, or, alternatively, there has been a huge error in the integration. From [3] we know the answer shoudl be in the end:

$$W = C\frac{me^4}{\hbar^3}$$

Is our answer in this form, is a good question to ask? Well... we can convert k into its definition.

$$ W = \frac{2^{16}\pi^{9/2}m^{1/2}e^4}{3^2\hbar^2a^9\sqrt{-E_{He}}} | IntegralInterior |^2 $$

And the energy can be converted as well. 

$$ W = \frac{\sqrt{2}2^{16}\pi^{9/2}me^4}{3^2\hbar^3a^8} | IntegralInterior |^2 $$

And that is in fact of the right form, $\frac{me^4}{\hbar^3}$. Naturally, we expect the IntegralInterior to evaluate to something of the order of $a^8$, which... well, it certailny will produce terms *of* $a^8$ but it may also produce smaller order terms, we would have to evaluate it out completley to see. (Chances are there's an algebra mistake in there that leaves in some trailing terms.)

While this is definitely a mess and not satisfactory, at the very least this work has shown that we are on the right path. The form is correct and the final evaluation looks hopeful, if tedious.
