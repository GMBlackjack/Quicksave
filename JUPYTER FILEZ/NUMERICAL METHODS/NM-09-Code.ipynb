```
#include <stdio.h>
#include <stdlib.h>
#include <math.h>

int counter = 0;

double f(double x,double x_min,double x_max) {
    ++counter; 
    double buffer = sin(1.0/x) - 1.0;
    double relErr = (x_max-x_min) / ((fabs(x_max) + fabs(x_min))/2.0);;
    printf("%i\t %8.7e\t %8.7e\t %8.7e\t %16.15e\n",counter, x, x_max-x_min, relErr, buffer);;
    return buffer;
}

int main() {
    double a=0.15;
    double b=0.5;
    double c = (a+b)/2; //the middle is rquried. c for "center." 
    
    double minValue = f(a,a,b);
    double maxValue = f(b,a,b);
    double midValue = f(c,a,b);
    //values used to store previous evaluations through iterations.
    //Initial setup requires three evaluations. 
    
    if (midValue > maxValue || midValue > minValue) {
        printf("ABORT! ABORT! ABORT!");
    }
    
    while ((b-a) / ((fabs(b) + fabs(a))/2.0) >= 1e-6) {
        double leftValue = f((a+c)/2.0,a,b);
        double rightValue = f((b+c)/2.0,a,b);

        if (leftValue < midValue) {
            b = c;
            maxValue = midValue;
            c = (a+c)/2.0;
            midValue = leftValue;
        } else if (rightValue < midValue) {
            a = c;
            minValue = midValue;
            c = (b+c)/2.0;
            midValue = rightValue;
        } else {
            a = (a+c)/2.0;
            minValue = leftValue;
            b = (b+c)/2.0;
            maxValue = rightValue;
        }
    }
    
    printf("FINAL: %16.15e\t %16.15e\n", c, midValue);

    return 0;
}
```
