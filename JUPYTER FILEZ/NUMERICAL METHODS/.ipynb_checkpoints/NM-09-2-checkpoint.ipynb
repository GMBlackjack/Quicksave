Here's the pseudocode for the Golden Section Search.

```
counter = 0

f(x,x_min,x_max)
    ++counter 
    buffer = sin(1/x) - 1
    relErr = (x_max-x_min) / ((|x_max| + |x_min|)/2)
    print(counter, x, x_max-x_min, relErr, buffer)
    return buffer

main()
    phi = (sqrt(5) - 1)/2
    
    a=0.15
    b=0.5
    c = b + phi*(a-b)

    minValue = f(a,a,b)
    maxValue = f(b,a,b)
    midValue = f(c,a,b)
    //values used to store previous evaluations through iterations.
    //Initial setup requires three evaluations. 

    if !(midValue < maxValue && midValue < minValue) 
         error //you chose the wrong type of points for this! 
        
    d = a + phi*(b-a) 
    newValue = f(d,a,b) // the test point needs to be set once outside the loop. 

    while (b-a) / ((|b| + |a|)/2.0) >= 1e-6

        if (midValue < newValue)
            b = d
            maxValue = newValue
            d = c
            newValue = midValue
            c = b + phi*(a-b)
            midValue = f(c,a,b)
        else
            a = c
            minValue = midValue
            c = d
            midValue = newValue
            d = a + phi*(b-a)
            newValue = f(d,a,b)

    print(c, midValue)
```

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

First, let's calculate some derivatives. The function in question is

$$ sin(1/x) - 1 $$

We can take the derivative using the chain rule.

$$ - cos(1/x) \frac{1}{x^2} $$

And for the second derivative the product rule.

$$ - sin(1/x) \frac{1}{x^4} + 2 cos(1/x) \frac{1}{x^3} $$

And this is the second derivative we need.

Now, we must ask what $\alpha$ is? It is the point near the minimum that our program finds. We want it to six sig figs, so we end up with $\alpha = 2.12207$ (six sig figs is the same for both methods so it matters not which program we got it from). So what we're evaluating is..

$$ \sqrt{\frac{|f(\alpha)|}{\alpha^2f''(\alpha)}} = \sqrt{\frac{|sin(1/\alpha) - 1|}{\alpha^2(- sin(1/\alpha) \frac{1}{\alpha^4} + 2 cos(1/\alpha) \frac{1}{\alpha^3})}} $$

At this point we turn to the computer to determine what the value actually is. We get...

$$ x_c / 2.12207 \approx 0.783825 $$

That is within an order of magnitude of 1 so we meet the criteria we were given.

# Problem 4 \[Back to [top](#toc)\]
$$\label{P4}$$

This is some pretty basic probability theory. First, the expected number of total cycles is equal to the expected number of cycles for one iteration times the number of iterations. So let's consider just one iteration at a time. 

When we have N nested if statements, all of them have 1/N probability. This is a Uniform Probability Distribution, which means the expected value is precisely halway between the minimum clock cycles and maximum clock cycles every time. So, considering only the if statements, the expected cycles are always (max - min)/2. To get the total for the iteration we then add the cycles for the funciton evaluation to this, since we do it every time.

So, let us do this for N=2.The minimum number of cycles for "if" is 20, while the maximum is 40, so we expect 30 to come from if cycles. Each iteration also has a single "do something" increasing the value to 32 per iteration. Since MMA requires 160 iterations, we multiply the numbers to get 5120 cycles. But we aren't done yet, as we also have to add the function evaluations, 320 evaluations times 10 cycles is 3200. Thus, the total expected value for MMA is 8320.

For MMB, we do much the same. 20 minimum if cycles, 60 maximum since we have N=3, making the expected value 40. Add 2 for the single "do something" and we end up with 42 expected. Times 100 for the iterations we get 4200. Function evaluations are 100 at 10 each so they provide 1000, giving us a grand total of 4200 + 1000 = 4200. 

Note that 8320 > 4200, so MMA is more expensive to run. despite only having two if statements.
