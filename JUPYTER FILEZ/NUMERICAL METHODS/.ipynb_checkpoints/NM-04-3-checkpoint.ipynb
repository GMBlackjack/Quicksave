$$ \text{ for } n = 0,1,2,3...$$

$$ x_0 = 0: 0, 1, 0, -1, ... $$
$$ x_0 = \pi/4 : 1/\sqrt{2}, 1\sqrt{2}, -1\sqrt{2}, -1\sqrt{2}, ... $$
$$ x_0 = \pi: 0, -1, 0, 1, ... $$
$$ x_0 = 3\pi/2 : -1, 0, 1, 0, ...$$

Now, we're told we can use eight points. If we wanted to be *extremely* accurate, we would use the symmetry of the basic trig functions to note that we only need to know poitns along a fourth of a wavelength (0 to $\pi/2$) and that all the others could be found by reflections. However, given that we are only given the square root of two, we should probably restrict ourselves to steps of $\pi/4$ in this program, in which case the eight points would be:

$$ 0, \pi/4, \pi/2, 3\pi/4, \pi, 5\pi/4, 3\pi/2, 7\pi/4 $$

And $2\pi$ can replace 0 at will. This, notably, onlyc overst he range 0 to $2\pi$, but using periodicity we can get it to solve values from -10 to 10. (In fact it will be easier to program this in complete generality so it can solve any range). 

Regardless, we have found points that were not noted in the first part of the problem. Fortunately using the magic of the trig identities we can adjust the patterns. (Or we could just plug in the x_0 values manually since we do have the generic form, but it's easier to do the trig adjustments in our head). Regardless, the results are...

$$ \text{ for } n = 0,1,2,3...$$
$$ x_0 : sinx_0, cosx_0, -sinx_0, -cosx_0, ... $$
$$ x_0 = \pi/2: 1, 0, -1, 0, ... $$
$$ x_0 = 3\pi/4 : 1/\sqrt{2}, -1\sqrt{2}, -1\sqrt{2}, 1\sqrt{2}, ... $$
$$ x_0 = 5\pi/4 : -1/\sqrt{2}, -1\sqrt{2}, 1\sqrt{2}, 1\sqrt{2}, ... $$
$$ x_0 = 7\pi/4 : -1/\sqrt{2}, 1\sqrt{2}, 1\sqrt{2}, -1\sqrt{2}, ... $$

Now we just program a funciton that takes an arbitrary x and gives us a proper value. We'll describe exactly how it works in part a). The code itself can be found in the coding section. Looking ahead, we also made the code able to spit out all the values from -10 to 10 in the requested range, but that's for a later part.

We specified the values of $x_0$ above as that needed to be decided before beginning programming. 

Since we could not assume we knew pi, we calcualted it using the second algorithm from Homework 1, since we could get maximum precision from it. 

Before anything, the program checks to see if the point is between 0 and $2\pi$. IF it is not, it adds or subtracts multiples of $2\pi$ until it is. Even though we only use -10 to 10 values here, the program could do however many is desired, though for very large ranges naturally repeated subtractions wouldn't be a prefered method. 

However, given an x, how is an $x_0$ chosen? Well, it just so happens that we selected equidistant $x_0$ points, separated by the same amount, and that separation also happens to be one of the values we used! $\pi/4$. We can use this number as a measuring stick to determine which $x_0$ x is the closest too--it'll be the single point with a distance to x that is less than $\pi/4$. Granted, this doesn't consider the case of a tie, but the program just assigns it to the lower value in this case. 

You may note that this leaves out the range of poitns near $2\pi$. in this case, the program makes a note that it needs to shift the compulation, but otherwise uses the values for $x_0 = 0$.

Once we have determined which point we are close to, we calculate what the result is by adding up the taylor series for the selected point from n=0 to n=12 exactly, including the zero terms just to be extra thorough. (And to follow the instructions). However, we did have to program a power function here. Wasn't difficult.

AFTER THIS, we added 2 to the result. Since the problem asks for sin(x)+2, not sin(x). This is presumably to make errors more well behaved. 

Before printing everything out, we go ahead and calculate the relative error. We noticed that a lot of the time the bulit in sin(x) function and the program we just made agreed perfectly, getting errors of -inf. This is ridiculous so whenever the result appeared to be exact, we just said the error was -16, as reasonably the machine wasn't giving us any more digits than that. This keeps the future plots from being misleading--the error is not *actually* zero, the system itself only reports up to a certain precision.

(There is also a check here to make usre the error doesn't rise above -14. It never did.)  

So now everything gets printed to the file, and then we're done!
