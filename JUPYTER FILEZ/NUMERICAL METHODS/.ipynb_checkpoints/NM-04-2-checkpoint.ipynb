# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

For the record the two rules are "no complicated funcitons." This includes the factorial function. This doesn't seem like two different points to me, but they are listed separately.

Anyway, every term in the Taylor series is written like:

$$ \frac{f^{(n)}(x_0)}{n!} (x-x_0)^n $$

However the problem asks specifically for the coefficients of just $f^{(n)}(x_0)$, so we shall leave off the other considerations for now. 

The derivatives of sine follow a cycle: $sinx \rightarrow cosx \rightarrow -sinx \rightarrow -cosx \rightarrow sinx$. This can be used to formulate a pattern for all derivatives at any value $x_0$. We are asked specifically for $x_0 = 0, \pi/4, \pi, 3\pi/2$. Since we always hav ea cycle, we can define the pattern as a cycle of 4 no matter what $x_0$ is.

$$ \text{ for } n = 0,1,2,3...$$
$$ x_0 : sinx_0, cosx_0, -sinx_0, -cosx_0, ... $$
$$ x_0 = 0: 0, 1, 0, -1, ... $$
$$ x_0 = \pi/4 : 1/\sqrt{2}, 1\sqrt{2}, -1\sqrt{2}, -1\sqrt{2}, ... $$
$$ x_0 = \pi: 0, -1, 0, 1, ... $$
$$ x_0 = 3\pi/2 : -1, 0, 1, 0, ...$$

Now we convert this to code which is on the Code part of this assignment, not reproduced here. We will take in a value n and a value $x_0$ and then release the coefficient of the pattern, for any n from 0 to arbitrary int of whatever size the compiler can handle. We are allowed to use modulo, but we are NOT allowed to use square root. Fortunately we are provided by the problem (in part 2's bullet points) an approximation $\sqrt{2} \approx 1.4142135623730950488$.

We were tempted to write out entire algorithms from scratch to calculate the sine and cosine functions as exponentials, but that would be a ton of effort and is definitely not worth it, so the program will only accept the four given values of $x_0$. 

As of now there is no real output to report, the program just takes in $x_0$ and n and spits out the result. 

For this we will need a factorial funciton and an absolute value function.

Fortunately from previous homework we already made a factorial function.

The absolute value function can be accomplished by a simple "if negative, multiply by -1" function. 

After that, we wrote a program using all the functions to do as asked. The code itself is in the Code portion of the assignment. The *print* result, however, is reported here as the answer to the problem.

Note that all the zero coefficients are suppressed, but it is possible to imply their existence by holes in the reported n values. Also note that the signs of the results cycle just as the patterns for the first part of the problem. Also note that we hit 1e-31 at about the same n-step for each result. This is because the n! term largely determines coefficient size.
