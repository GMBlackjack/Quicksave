## Problem 1

*a) Given an arbitrary set of numbers {$ M_{\alpha\beta}; \alpha=0,1,2,3; \beta=0,1,2,3$} and two arbitrary sets of vector components {$ A^\mu, \mu=0,1,2,3 $} and {$ B^\nu, \mu=0,1,2,3 $}, show that the two expressions* 

$$ M_{\alpha\beta} A^\alpha B^\beta = \sum^3_{\alpha=0}\sum^3_{\beta=0} M_{\alpha\beta} A^\alpha B^\beta $$

*and*

$$ \sum^3_{\alpha=0} M_{\alpha\alpha} A^\alpha B^\alpha $$

*are not equivalent.*

This is trivial and can be shown by counterexample. Let $M_{10}$ be 1 and every other M-number 0. In the bottom sum, the result is zero, but in the top sum, there will be a single result. (Fun fact: we can also come up with specific cases where it is true, such as when only $M_{00}$ had a nonzero value.)

*b) Show that*

$$A^\alpha B^\beta \eta_{\alpha\beta} = -A^0B^0+A^1B^1+A^2B^2+A^3B^3$$

Trivially so by the definition of the metric tensor. -1 at 00, 1 at 11, 22, 33, and 0 everywhere else. 

I think the point that is trying to be shown here is that the metric tensor is not the same as multiplying everything together, I guess. It takes the place of the M and is one of the cases where the two of them are equivalent.
