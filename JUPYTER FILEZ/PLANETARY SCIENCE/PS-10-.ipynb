# Problem 10.1 \[Back to [top](#toc)\]
$$\label{P10.1}$$

*Estimate the largest negative value of $1/a_0$ that nongravitational forces are reasonably likely to produce for a comet that is initially in heliocentric orbit and is not subjected to significant planetary perturbations. You may assume that the comet's initial orbit is so eccentric that it may be taken as parabolic and that the comet outgasses 0.1% of its mass with a 10% asymmetry and that this outgassing occurs in a single burst at perihelion, which is at 0.2AU.*

So, outgassing, let's be very generic and liberal with this since the actual full equation is a nightmare to work with. The actual mass of our comet is so small compared to the sun that it doesn't matter what the total before and after is, it might as well just be M. However, it is important that the ejected amount m is 0.1% of M. Now, by outgassing we outgass in all directions, so the 10% assymetry must refer to the fact that 10% of the "force" this mass imparts is going to go into actually accelerating the comet, so now we're at 0.01$ of actual force involved.

The trick now is what is the acceelration of this dust? If we can get that, we can get the final result. Some searching suggests that some particles move with the Solar Wind. Since we're looking for MAXIMUM possible displacement, let's make these things book at at a solar wind maximum of 750 km/s. At perihelion the tail is flowing away from the sun, so the comet itself is pushed forward. So, overall, the energy change is:

$$ \frac12 0.0001 M (750000)^2 $$

However we do need to consider the direction this energy is being "applied". The push will move it closer to the sun! And would be the same as suddenly increasing the gravity for a moment, which shrinks orbits.

Okay so perhaps the tail isn't the source of our burst. instead, let's have our burst favorably be in the opposite direction of motion, thus boosting the comet forward! However this does mean that our energy calculation no longer works as we don't have a known speed. And if we were to treat it like a rocket we would also need a rough relative speed calculation. 

Let's try setting up conservation of momentum.

$$ Mv_i = (M-m)v_1 - mv_2 $$
$$ \Rightarrow Mv_i = 0.9999Mv_1 - 0.0001Mv_2 $$
$$ \Rightarrow v_i = 0.9999v_1 - 0.0001v_2 $$

Now we can find $v_i$ from the parabolic trajectory, but the other two not so much, as the two different velocities are independent. 

There has to be some way to find the ejection speed of the comet's mateiral... aha! It ejects at 1/2 km/s ISH near 1 AU, but then it expands and speeds up to up to 2km/s. This is Haley's Comet's numbers, so we can probably use them.

Now the velocity of a prabolic orbit is $\sqrt{2GM/r}$, which we can calculate out to be 42.1143 km/s. So the improved speed is now... 42.1187 km/s!

Now we just need to associate this to a hyperbolic orbit. The equation for 1/a is...

$$ 1/a = 2/r-v^2/GM $$

1/a = -2.79e-15 1/m

Woo-ee that's a small difference. BUUUT it's in inverse meters! We need to convert to inverse AU, which gives us -4.191e-4 1/AU. AS we were expecting values up to -1e-3 this is a perfectly acceptable answer. (In figure 10.3 -0.001 appaers to be the far left of the diagram, and there are no comets there. There are a few at what appear to be -0.00075, but the vast majority are within -0.0005, which is about where our comet lies. A particuarly drastic breakup could cause the few more hyperbolic orbits.)

Keep in mind we were using Halley numbers, which only has 0.6 AU perihelion!

# Problem 10.4 \[Back to [top](#toc)\]
$$\label{P10.4}$$

*Estimate the frequency at which we would expect to observe interstellar comets. Assume that the Oort cloud contains 1e12 comets that are large enough to be observable if they were to reach the inner solar system, and that five times as many comets have been ejected from the Solar System over time. Postulate that the number of comets produced and released by a planetary system is proportional to the star's mass and that our Solar System is average in this regard. Furthermore, assume that the density of stellar mass in the solar neighborhood is 0.065 earth masses per cubic parsec. (1 parsec is 3.1e18 cm). The typical velocity of a comet relative to the sun at infinity is comparable to that of neighboring stars, 30km/s. Also assume that a coment must come within 2 AU of the Sun to be visible.*

*a) Neglecting gravitational focusing by the Sun, i.e., approximating the trajectories of the comets by straight lines.*

Let's ignore the sun itself and just pretend we are trying to hit at disc that is 2 AU in diameter, so we have a "cross sectional area" of 12.566 square AU. 

So, let us say that all "ejected" comets are distributed randomly more or less regarldess of which star they came from. The sun has released 5e12. If ejection number is proportional to mass, this allows us to find the constant of proportionality: 1.989e30 = A5e12, the constant A is 3.978e17 kg/comet, or 2.514e-18 comet/kg. 

We were considering just using the number of stars of various colors and integrating to get total mass, but rather than that we've just been given the overall density in stellar mass, 0.065 earths per cubic parsec. This is 3.88e23 kg per cubic parsec. Using our constant of poportionality, we end up with 9.754e5 comets per cubic parsec. These are specifically ejected comets spinning around randomly. 

If the average speed is 30km/s, every comet will cross the length of the cubic parsec in about 1.033e12 seconds. We could also write this as 32,000 years. 

So if 9.754e5 comets cross the length of the cubic parsec in 32,000 years, how many of them hit our circle? Well, this is actually just a case of converting from 12.566 square AU to a square parsec, 2.954e-10. This is 2.881e-4 comets. You heard me, less than *one* every 32,000 years. This means we don't expect the first comet for 111 million years! 

*b) Including gravitaitonal focusing by the Sun, i.e., considering comets to be traveling on hyperbolic orbits around the sun.*

Okay so this here is going to require finding the cross sectional area of hyperbolic orbits, and the result is that the square AU conversion will change, all other numbers remain constant. This means we need to know the angle of scattering in a hyperbolic orbit that just barely reaches 2AU. This means we're going to need the angle of the hyperbolic asymptote. We have ways to find the semimajor axis a, in fact if we have the same velocity at infinity there's no ambiguity. For 30km/s, a=0.986 AU. 

The problem is, a is not sufficient to define the asymptote, we also need something else. 

The periapsis distance is given as -a(e-1), which means e is our parameter we can specify. For a "barley scratching" orbit we have e = 3.028 which, yeah, tracks for a highly hyperbolic orbit. 

The semi-minor axis (which we will use to characterize the asymptote) is $b = - a\sqrt{e^2-1}$. With our values b=2.19 AU. 

Now, with this we are armed to find the "radius of cross section," which can basically be thought of the distance of closest approach by the asymptote. The line can be represented by a slope of |b|/|a|. Our goal is to find the point on this line that is closest to the position of the sun $|a|+r_p$, where $r_p$ is the periapsis distance. 

So let's generate a function for this, as we will have to do it again for part c). Our asymptote is $y = |b/a| x$. The function we want though is the distance from any point x to the point $(0,|a|+r_p)$. This distance will always be

$$ \sqrt{(|a|+r_p-x)^2 + y^2} = \sqrt{(|a|+r_p-x)^2 + (|b/a|x)^2} $$

Now as finding the minimum of the square is sufficient, we can get rid of the root. 

$$ (|a|+r_p-x)^2 + (|b/a|x)^2 $$

This is what we need to minimize. With constants set to 1, the shape is:



Which we can see *has* a minimum so this should be doable. So we take the derivative:

$$ -2(|a|+r_p-x) + 2(|b/a|^2x) $$

Now the minimum is when the derivative equals zero.

$$ \Rightarrow -|a|-r_p-x + |b/a|^2x = 0$$
$$ \Rightarrow (|b/a|^2-1)x = |a| + r_p$$
$$ \Rightarrow x = \frac{|a| + r_p}{|b/a|^2-1}$$

As long as |b| > |a| this produces a positive value. Which is what we have. If |a|>|b| the equation breaks. (Note that |a| = |b| the closest point bisects (0,0) and the point directly above the sun) Fortunately for us our |a| > |b| so we're completely safe! 

So running the calculation gives us a distance of... 0.416 AU. Note that this is the distance from the origin, not the sun, the actual distance from the sun here is 2.570 AU. 

Which is to say the sun isn't focusing things *very* much in terms of scattering area. But this still increases the area to 20.746 square AU. This is 4.876e-10 square parsecs. This translates to 4.756e-4 comets per 32,000 years. We should see one every 67 million years in this case. 

*c) Repeat parts a) and b) assuming that we can observe all comets that approach to witihn 5 AU of the sun.*

This is a bit of wishful thinking, comets don't always react at 5 AU, so not all of them will "comet up" at this point. But oh well, let's go ahead and extend the distance. 

For part a) this just gives us a cross sectional area of 78.540 square AU, or in square parsecs, 1.846e9. A mere 17 million year wait! 

Now if we allow the sun to grab stuff, we need to get ourselves a new b. The closest approach is 5 AU so... b = 5.904 AU. This gives us "x=0.141" which corresponds to a distance of 5.845 AU, thus the cross section is 107.313 square AU, or in square parsecs, 2.522e-9. This reduces the wait time to 13 million years! 

Basically we shouldn't expect an extraterrestrial comet ever. 

In fact our numbers are so low we're a little suspicious... we *have* seen a single interstellar object, and it passed through *Mercury's* orbit. So clearly these things *do* show up.

# Problem 10.9 \[Back to [top](#toc)\]
$$\label{P10.9}$$

*An asteroid and a comet have the same apparent brightness while both are at $r_\Delta = 2AU$ and $r_\odot = 3AU.$ At a time later they are both observed at $r_\Delta = 2AU$ and $r_\odot = 2AU$. Which object is brighter and by approximately how much?*

The comet, obviously, it's far more reflective. Though the "how much" question, well... let's do some thinking. 

Okay so there's some complexities here that have now come to light. The actual abledo of comets is very very dark, but this is just the *nucleus*, within the 2-3AU distances the comet's brightness is primarily determined by the coma. 

Since the earth distance is the same in both cases, it is irrelevant to the relative brightnesses of the two objects as they both will dissipate $1/r^2$ like everything does. 

Turns out we have a relative brightness equation, 10.3 $B \propto \frac{1}{r_\odot^\zeta r_\Delta^2}$. Asteroids have $\zeta=2$ while comets, due to becoming more active closer to the sun, have larger values. The exact value of $\zeta$ appears to be variable for comets, so we shall treat it as a known constant but give it no numbers. WE already discussed the earth distance being irrelevant, so...

$$ A \frac{1}{3^\zeta} = A' \frac{1}{3^2} $$

This is for when they were both at 3 AU. Note the proportionality constant, which is unfortunately not the same for both as then it would require both $\zeta$ values to be the same. 

Now, what are the adjustments for each? Well..

$$ A \frac{1}{2^\zeta} $$ 
$$ A' \frac{1}{2^2} $$

The asteroid adjusts from 1/9 to 1/4, a factor of 9/4 brighter, so 2.25. 

The difference between $1/3^\zeta$ and $1/2^\zeta$ is a bit harder to define.

$$ \frac{3^\zeta}{2^\zeta} $$

So naturally for all $\zeta>2$ this is more than the asteroid. The *exact* value of proportionality between them woudl be

$$ \frac{4}{9}\frac{3^\zeta}{2^\zeta} = \frac{3^{\zeta-2}}{2^{\zeta-2}} $$

# Problem 10.12 \[Back to [top](#toc)\]
$$\label{P10.12}$$

*At a distance $r_\odot$ = 1 AU, outgassing is dominated by H2O and a comet's surface temperature is 190K. The latent heat of sublimation is $L_s \approx 5e11 erg/mol$. Assume the bond albedo $A_b=0.1$ and the infrared emissivity $\epsilon_{ir} = 0.9.$*

*a) Derive equation 10.6 and use it to determine the gas production rate of this comet.*

The equation is:

$$ Q \approx \frac{(1.2e18)\pi R^2}{r_{\odot AU}^2} $$

And we are told it was derived with a bond albedo of 0.1, which we are in fact assuming. Q is the gas production rate. 

This... is looking to be a pain, because the equation to derive it from is:

$$ (1-A_b) \frac{\mathcal F_\odot e^{-\tau} \pi R^2}{r^2_{\odot AU}} = 4\pi R^2 \epsilon_{ir} \sigma T^4 + Q\frac{L_s}{N_a} + 4\pi R^2 K_T \frac{\partial T}{\partial z} $$

The last term is small for comets and usually ignored, so...

$$ \Rightarrow (1-A_b) \frac{\mathcal F_\odot e^{-\tau} \pi R^2}{r^2_{\odot AU}} = 4\pi R^2 \epsilon_{ir} \sigma T^4 + Q\frac{L_s}{N_a} $$

Solve for Q so we can have some idea what we're actually going for.

$$ \Rightarrow (1-A_b) \frac{\mathcal F_\odot e^{-\tau} \pi R^2}{r^2_{\odot AU}}\frac{N_a}{L_s} - 4\pi R^2 \epsilon_{ir} \sigma T^4\frac{N_a}{L_s} = Q$$

MOST of what we have here are numbers that are provided or obvious. Those that aren't are R, r, and $\tau$. Of these only $\tau$ is a problem as we need to get rid of it in the final expression. We'll see what happens. first. Just evaluate things...

Man this is a units mess waiting to happen, note how the first term is over AU, but the $\sigma$ constant most definitely is not in terms of AU. This'll be... annoying to be sure. 

Bizarrely the units don't seem to match, we have $W/AU^2$ on one and just @ on the other. (ignoring the N/L factor for the moment). We can justify this as the AU radius as just a scaling factor for the flux, however unless we can justify ignoring the second term it would demand an AU-based section in the numerator, which is not in what we have provided as the answer by the problem. 

Still, we can evaluate with various conversions to get...

$$ \Rightarrow \frac{ 1.640e22 e^{-\tau} \pi R^2}{r^2_{\odot AU}} - \pi R^2 3.204e21 = Q$$

Second term is annoyingly not neglegible, and does not have the radius factor on it. Joy.

Now, what is the optical depth? Google search suggests that at the surface it reaches about 10. So...

$$ \Rightarrow \frac{ 7.446e17 e^{-\tau} \pi R^2}{r^2_{\odot AU}} - \pi R^2 3.204e21 = Q$$

Now we note that the first term is *close* to what we want, possibly just due to an argument over what the optical depth actually is. But in that case, why would we ignore the second MUCH LARGER term? The only thing we can think of is the scaling difference but we can't see how that would change anything if the untis agree. If they don't agree, how are we even adding them...?


Regardless we can still USE the given version. Rather trivially we end up with 

$$ Q \approx (3.770e18) R^2 $$

Note that nowhere are we told what the radius of this comet is, though in part b) we are told it ranges from 1 to 10 km. If we take this we get a range of values:

$$ 3.770e24 < Q < 3.770e26 $$

In molecules per second. 

*b) Calculate the thermal expansion velocity of the gas and compare this to the escape velocity of a typical comet, radius 1-10 km.*

Thermal expansion velocity is given by the relation:

$$ \frac12 \mu_a m_{amu} v_0^2 = \frac32 kT $$

Which those of us in the know can recognize as a form of the Equipartition Theorem. Just replace the unusual variables with mass and get

$$ v_0^2 = \frac{3kT}{m} $$

The mass in this case is that for water, 2.989e-26 kg. The result is a velocity of 513 m/s. Speedy! Nowhere near enough to escape from EARTH... but what about a comet?

Rather than just looking it up let's actually do it. Consider a comet to be a 10km sphere made out of water, so a total mass of 4.189e15 kg. Escape velocity is $\sqrt{2GM/r}$ as usual, which means... 7.477 m/s. The outgassing *easily* escapes, which is probably why the tail is so *massive*. 

This does match what we know, Haley's Comet has an escape velocity of 2 m/s. *two*. Fun. 

Granted the gas might not shoot off at directly the right direction, but still 513 is high enough that basically everything escapes. Explains why dust can be dragged away as well.
