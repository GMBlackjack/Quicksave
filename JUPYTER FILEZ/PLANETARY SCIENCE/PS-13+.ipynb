*Consider an H2 molecule that falls from infinity to a circular orbit at 1 AU from a 1 $M_{\odot}$ star.*

*a) Calculate the circular velocity at 1 AU, and determine the total mechanical (kinetic + potential) energy of a molecule on a circular orbit at 1 AU. Note that the total energy of the molecule at rest is initially zero.*

OKAY we were caught by the trap. One can't just set kinetic energy equal to gravitaitonal energy, that doesn't work, while total energy must be conserved overall, something has to transform the energy to get it all to move into an inner orbit. 

$$ v = \sqrt{\frac{GM}{r}} $$

Which, at 1 AU, is Earth's speed of 29.8 km/s. Place it into $\frac12 mv^2$

Now this is the kinetic term. The potential term is our usual -GMm/r. However, if we try to set the kinetic and gravitaitonal energies equal: 

$$ \frac12 mv^2 = \frac{GMm}{r} $$
$$ \frac12 v^2 = \frac{GM}{r} $$
$$ v = \sqrt{\frac{2GM}{r}} $$

This isn't right! That means there's extra energy lying around that had to go somewhere. There is another factor of $\frac12 mv^2$ in there. 

$$ E = \frac12 mv^2 + \frac12 mv^2 - \frac{GMm}{r} $$

Now this absolute total energy has to go to zero. But why is the kinetic energy doubled? Ah, well, that energy goes into other things. Part b) suggests that we should consider it TEMPREATURE INCREASE. 

*b) Calculate the temperature increase of the hydrogen gas assuming it has not suffered any radiative losses.*

For each individual particle by the equipartition theorem:

$$ E = \frac32 kT $$

Mutliply by N to get the total energy.

$$ E = \frac32 NkT $$

Of course we do only care about single particles so...

$$ T = \frac{mv^2}{3k} $$

In our case speed is 30km/s, mass is 2amu, and k is the boltzmann constant. After converting to SI we get

T = 7e4 K.

This is exactly what the book suggests we should get. That said, this temperature is not attained, the timescale for infalling allows for heat radiation to occur.

# Problem 13.4 \[Back to [top](#toc)\]
$$\label{P13.4}$$

*Show that the density of gas in a thin isothermal circumstellar disk varies in the direction perpendicular to the midplane of the disk as*

$$ \rho = \rho_0 e^{-z^2/H_z^2} $$

*where the Gaissuan scale height is given by 13.11. Hint: consider a balance between pressure and the component of the star's gravity perpendicular to the midplane of the disk.*

For the record:

$$ H_z = \sqrt{\frac{2kTr_\odot^3}{m GM_\odot}} $$

Anyway, we have two forces at play here. A pressure differential resulting in a net force, and the force from gravity. 

Force from gravity is pretty easy on any individual particle: $\frac{GMm}{r^2} sin\theta $, though in our case we definitely have small angles as this is a *disc* and thus this might as well just be $\frac{GMm\theta}{r^2}$. This pulls downward.

The question now is what is the force on the particle? A quick lookup provides this equation:

$$ F = -\frac{m}{\rho} \frac{dP}{dz} $$

So balancing everything out provides us with

$$ \frac{GM\theta}{r^2} = -\frac{1}{\rho} \frac{dP}{dz} $$

We'll want everything in terms of z. In which case, we note that $z = rsin\theta$, so if we substitute we get:

$$ \frac{GMz}{r^3} = -\frac{1}{\rho} \frac{dP}{dz} $$

So now we can rearrange and get a somewhat simple differential equation.

$$ -\frac{GM\rho}{r^3}z = \frac{dP}{dz} $$

Now while one might *think* we can get this by a simple integration, remember that $\rho$ varies with z as well. One thing we do know is that $P \propto \rho$ for a gas. ($P = \frac1\mu \rho kT$ for an ideal gas, where $\mu$ is individual molecular mass). 

$$ -\frac{GM\mu}{kTr^3}zP = \frac{dP}{dz} $$

The solution to this differential equation is going to be a square exponential. We ignore the positive exponential because it explodes at large distances, which doesn't make sense on physical grounds. The answer is then of the form:

$$ P = Ae^{-ax^2} $$

Now given our solution we can assume we have initial conditions such that we can set $A = P_0$. a, meanwhile, is a constant: $\frac{GM\mu}{2kTr^3}$

$$ P = Ae^{-\frac{GM\mu}{2kTr^3}x^2} $$

Note that if we flip the fraction and square root it we end up with H.

*a) Calculate the amount of gas that a particle R cm in radius orbiting at 1 AU from a 1 solar mass star passes through (collides with) during one year. You may assume that the density of the protoplanetary disk is 1e-9 $g/cm^3$ and $\eta$ = 5e-3.*

This could be approximated using cross sectional area... or we could just use the torus volume formula.

$$ 2\pi^2 r^2 R $$

R is the bigger one, aka 1 AU. Since we have density... wait hold on what's $\eta$? It doesn't seem like it would be necessary to find the amount of particles, we have the density. 

AHA! dust doesn't rotate at the Keplerian Velocity! And if it did we wouldn't run into it! THERE'S the issue! The angular velocity of gas is

$$ n_{gas} \approx \sqrt{\frac{GM}{r^3}}(1-\eta)$$

Which naturally coresponds to a tangential velocity of

$$ v_{gas} \approx R\sqrt{\frac{GM}{r^3}}(1-\eta)$$

Where "1" is the standard, so as far as our particle is concerned these particles are moving a factor of $\eta$ slower than it. So in the particle's frame...

$$ v_{gas} -R\approx \sqrt{\frac{GM}{r^3}}(\eta)$$ 

Since they're orbiting the same direction this will result in a dramatic decrease in observed "density". So we decide *not* to solve this in the torus way but in the 1D way, treating the particle still and just using timing. It takes 1 year to go around at 1 AU, so in order to get a distance traveled we simply take tv, that is velocity times time. With the distance (which is relative) we then proceed to multiply it by the cross sectional area $\pi r^2$. Thus our full answer is the form:

$$ \rho \pi r^2 t R\sqrt{\frac{GM}{R^3}}(\eta) $$

Note that we have suppressed the sign for our sanity. Anyway we now have all the values we need. Plugging everything in and calculating gives us...

$$ = \rho \pi r^2 t \sqrt{\frac{GM}{R}}(\eta) $$
$$ = \rho \pi r^2 t (30000 m/s)(\eta) $$
$$ = \rho \pi r^2 t (150 m/s) $$
$$ = \rho \pi r^2 (4.730e9 m) $$
$$ = 1486 r^2 $$

Tah-dah! Since we used R for the large radius, the particle radius is r.

Just set accumulated and particle mass equal to each other...

$$ \Rightarrow 1486 = 12566 r$$
$$ \Rightarrow 0.118 = r$$

This is in meters. So about 11.8 cm. This seems reasonalbe actually.

# Problem 13.19 \[Back to [top](#toc)\]
$$\label{P13.19}$$

*Consider a simple model for the accretion of the Earth. Suppose the radius of proto-Earth increases linearly with time from some starting point until accretion ends $t_{acc}$ years later, i.e., $R(t) = (t/t_{acc})R_\Phi$. Ignore the insulating effects of large impacts (which bury hot ejecta) and any possible atmosphere (which can prevent the surface from radiating freely to space). For further simplicity neglect compressibility, heat conduction, and internal heat sources, and set the emissivity to $\epsilon = 1$. The energy balance per unit area at the surface of the growing Earth is then quantified by Equation 6.27*

*a) Describe physically each of the three terms in Equation 6.27.*

And so begins the gauntlet. 

$$ \frac{GM(R)\rho}{R} \frac{dR}{dt} = \sigma (T(R)^4 - T_0^4) + \rho c_p (T(R) - T_0) \frac{dR}{dt}$$

So, a "simple" differential equation. The individual components aren't too difficult, we have initial temperature, temperature at a given R--but in this case R is the exterior radius, the surface, and it changes with time. The mass M is also changing with R, making a rather complex relationship. Naturaly in terms of accretion, dR/dt is a positive quantity that keeps increasing. 

So, what do each of the terms mean?

The left hand side is the gravitational energy available for heating. The actual energy *increase* is equal to the right-most term, which is directly related to $\Delta T$. What is the middle term? It's heat radiated away. But why isn't it negative then? Well, because it's on the wrong side for that! The heat radiated away is taken away from the total energy provided by gravitational collapse. Note that annoyingly it does require knowing what T(R) is at any given time so, uh, yeah quite annoying. 

Anyway that's what the equation here IS: total energy available = radiated heat + interior energy remaining.

*b) Assume the accreting material has an initial temperature $T_0 = 300K$, and that the solid particles have an average density $\rho = 4.5 g/cm^3$ and heat capacity $c_p = 1e7 erg/gK$. Find the approximate temperature of the planet as a function of radius assuming dR/dt = constant, for $t_{acc} = 1e8 yr$ and for $t_{acc} = 1e6 yr.$*

The annoying thing is that 6.27 only reports surface conditions. However, one thing we are told is to neglect a *lot* of things, including conduction. So anything not on the surface we consider to keep the temperature it had when it *was* on the surface, so once it's buried it's buried. Thus when we solve for a specific time t, we can also solve for any individual R by finding R(t). 

Given that dR/dt = constant, R(t) = At, with A being some constant. The constant *may* not matter, we'll see. 

Ignoring compressibility, M = $\rho \frac43 \pi R^3$. T(R) is currently unknown asn what we seek to find. So yes we do need to know A. However, we are actually told A in the form of R(t) = $R_\Phi t/t_{acc}$. $R_\Phi$ is the final radius at the end of accretion. Since we are dealing with proto-Earth, this can be assumed to be the Earth Radius, and $t_{acc}$ is a value we can freely set. With that, we are ready to restate our equation:

$$ \frac{4G\rho^2\pi R^2}{3} \frac{R_\Phi}{t_{acc}} = \sigma (T(R)^4 - T_0^4) + \rho c_p (T(R) - T_0) \frac{R_\Phi}{t_{acc}}$$

The annoying part is that what we need to solve for is T(R), and that's present in two places at different powers, and not ones with immediately obvious solutions. However, perhaps it will become clear that one term is far dominant over the other, and that it can be approximated away? Let's see, considering the left hand side only, $\sigma$ is the SB constant, and has an order of magnitude of 1e-8. (Specific value: 5.67e-8). This is essentially a change in the overall energy state. To figure this out, we need to find out what the gravitational binding energy of an object is. And since it's not evident the formula for this, we're gonna *make* it.

$$ dU = -G\frac{m_{shell} m_{interior}}{r} $$
$$\Rightarrow = -G \int_0^R \frac{m_{shell} m_{interior}}{r} dr$$

Now for something where we have a nice simple density, this goes like:

$$\Rightarrow = -G \int_0^R \frac{4\pi r^2 \rho \frac43 \pi r^3 \rho}{r} dr$$

Unfortunately our situation here is such that $\rho$ is "funky". For the uniform density distribution we don't have an issue:

$$ U = -\frac{3GM^2}{5R} $$
 
But *then* we have a differentiated earth. For the core this is no issue, we can model its gravitational potential easy enough:

$$ U_c = -\frac{3GM_c^2}{5R_c} $$

BUT the rest of the planet is now an ISSUE. 

$$\Rightarrow -G \int_{R_c}^{R_o} \frac{4\pi r^2 \rho_o \left[ \frac43 \pi R_c^3 \rho_c - \frac43 \pi R_c^3 \rho_o + \frac43 \pi r^3 \rho_o \right]}{r} dr$$

That "interior mass" is "mass of core minus mass of core if it were the smaller density plus mass of a sphere up to radius r of the smalle rdensity." Fun!

Anyway we can simplify a little bit.

$$\Rightarrow -G \frac{16}{3}\pi^2 \int_{R_c}^{R_o} \frac{ r^2 \rho_o \left[ R_c^3 \rho_c -  R_c^3 \rho_o +  r^3 \rho_o \right]}{r} dr$$
$$\Rightarrow -G \frac{16}{3}\pi^2 \rho_o \int_{R_c}^{R_o} R_c^3r \rho_c -  R_c^3r \rho_o +  r^4 \rho_o dr$$
$$\Rightarrow -G \frac{16}{3}\pi^2 \rho_o \left[ R_c^3 \rho_c \frac12 \left(R_o^2 - R_c^2 \right) -  R_c^3 \rho_o \frac12 \left(R_o^2 - R_c^2 \right)+  \rho_o \frac15\left(R_o^5 - R_c^5\right) \right]$$

So if we add this to $U_c$ we get the total binding energy of the core. Naturally, this is a huyge mess. And we generally can't just get a neat volume out of this, as the volume for a shell is $\frac43 \pi (R^3-r^3)$. We don't have the right powers and we can't extract them either as everything here is a difference.

The idea is simple. Take a pure gas, say hydrogen. What is the *smallest* coherent planet that can be made with it? (Least massive may also work, depends on which quantity is easier to work with). No rocky cores, no contaminants, and no bothering with thoughts of "well how could it form?" who cares, all I care is that it's stable. I'm hoping to show that you can have a coherent spherical cloud of hydrogen or something without any need to compress it into its metallic state...



Let us imagine a univesre where we add one hydrogen atom at a time, basically. One hydrogen atom doesn't count as a planet. If everything was at near 0K the atoms would not be moving at all and could accrete to each other in any number. If they were the only thigns in the universe ten hydrogen atoms at superchilled temperatures could hang out without a problem.

So okay fine a "handful of atoms" clearly isn't what we're looking for. So let's remove a restriction and say they have to be at least the background temperature of space itself, 3K, that there's some kind of cosmic microwave background they have to contend with. ...Actually hydrogen is solid at that temperature, let's up it to 10K. Thus we now have a speed:

$$ v_{rms} = \sqrt{\frac{3kT}{m}} $$

m is the mass of a single atom. So how fast *are* we going at 3K? The boltzmann constant k is 1.381e-23 J/K and the mass is 1.674e-27 kg. This gives us a rough speed of...

$$ v_{rms} = 497 m/s $$

Now, the question we next ask is: how much mass do we need to make that the escape velocity? Actually this is a bad question as this is rms speed, and we want very few particles escaping, so if we just had an escape velocity of 497 m/s we'd have an issue--namely losing mass as the random higher energy particles popped off. So let's double the energy. Since energy scales with $v^2$ we merely multiply our speed by $\sqrt 2$ to get 702 m/s. We shall consider this our necessary escape velocity.

$$ v_{esc} = \sqrt{\frac{2GM}{d}} $$

M is the mass of our ball of hydrogen, G is the gravitational constant, and d is the distance from it... or in our case, whatever we decide the radius is since the easiest place to escape is the surface. Problem is that, right now, both M and R are freely definable. 

Let's make an assumption: constant density. This will only be valid if the amount of mass needed to make our smallest planet is relatively small. If we run the numbers and get a big M out the end, then, uh, we'll have a problem. 

Anyway, if we have a constant density, than mass is related directly to radius via $\rho \frac43 \pi R^3$. And so...

$$ v_{esc} = \sqrt{G \rho \frac83 \pi R^2} = 2.364e-5 \sqrt\rho R $$

Now this is simpler to look at but we still have two freely specifiable quantities, R and the density. But we can actually plot them versus each other for an escape velocity of 702.

$$ \sqrt\rho R = 2.970e7 $$



On the x axis we have x in meters, but on the y axis we have the density in kg per cubic meter. The density of water, for instance, is 1000 $kg/m^3$. We may note that the denisties are much higher than this and, while hydrogen is compressible, this will make it so our planet isn't uniform if we try to use smaller and smaller radii. 

One thing we do need to do is determine what pressure corresponds to each density. We can use a variation on the ideal gas law for that:

$$ PM = \rho RT \Rightarrow P = \frac{\rho R T}{M}$$

M here is the molar mas, which is 1g/mol or 0.001kg/mol. 

$$ \Rightarrow P = 83140 \rho $$

You may note that presure is higher than density. So the densities on the graph are truly absurd. But now we can ask at what pressure does hydrogen stop behaving like gaseous hydrogen? At about 0.1 Pascals! This is a 1.2e-6 density object, which leads to a radius of 2.6e10 m, or 2.6e7 km. The radius of the sun is on the order 1e5 km, what we have created here is essentially a very loosely bound molecular cloud with no substance. 

This appears to be largely a consequence of our very low temperature. But this is just the boundary where the ideal gas law stops applying planet-wide... or cloud-wide. So it's unavoidable, we are going to have massive density differences. Which means we need a new strategy that's probably going to involve an Equation of State...

The accepted polytropic equation of state for gaseous planets is:

$$ P = K\rho^{5/3} $$

Problem here is that we don't have a limit on what K is. We'll have to poke around for reasonable values, it seems. We also need to contend with the pressure-gravity balance at such high pressures. What fun. 

Unfortunately we need to *find* K, as the provided K's we can find online are for the $\rho^2$ polytrope model. So how can we find a good K? Why, figuring out which value of K gives the best result for Jupiter, of course. Which means we're going to need to find hydrostatic equilibrium in order to find the mass with a given EoS. 

$$ \frac{dP}{dr} = -\frac{Gm\rho}{r^2} $$
$$ \frac{dm}{dr} = 4\pi r^2 \rho$$
$$ \rho = \left(\frac{P}{K}\right)^{3/5} $$

Combining these relations we can get a system of differential equations.

$$ \frac{dP}{dr} = -\frac{Gm}{r^2}\left(\frac{P}{K}\right)^{3/5} $$
$$ \frac{dm}{dr} = 4\pi r^2 \left(\frac{P}{K}\right)^{3/5}$$

We don't see a way to solve this analytically, so we are going to turn to a numerical solution. Fortunately this is just a simpler version of the TOV equations so we already have code to do it in OdieGM. Might be a *bit* annoying to get it to work, but we will. The value we are looking for is K, which is the one that gives us the Jupiter-like planet at the end of evaluation. 


Given huge uncertainties we were only able to get a rough boundary on K: between 1e5 and 1e6. 2.7e5 is what is recommended for the square polytrope for gas giants, however we are not using square polytrope, but the result does get some reasonable values. 1e6 provides a planet with larger radius but less mass than Jupiter which kind of makes sense given how we are using 100% hydrogen which likes to expand and is far less massive. However, the issue here is that we did use the Jupiter central pressure. This is the parameter we now have to adjust to see how small we can make a ball of hydrogen and still keep the atmosphere largely retained. We'll slowly lower the central pressure which will shrink M and R until we hit that point. 

$$ 497 = \sqrt{\frac{2GM}{R}} $$

We can actually resolve this somewhat, getting...

$$ \frac MR = 1.812e15 $$

So the Mass to Radius ratio needs to be this or higher to retain the hydrogen molecules. For Jupiter this ratio is 2.715e19, and for Saturn it is 9.759e18. This wouldn't really apply for Neptune and Uranus as they are not *mostly* hydrogen, but clearly both Jupiter and Saturn can hold onto their gas. 

Anyway, using both K 1e5 and 1e6 we will go until we find the result we desire.

!!!

Incredible.

So, an odd thing we noticed was that smaller central pressures led to LARGER radii, while more pressure led to SMALLER ones. We thought this was odd at first but we did remember that *most Jupiter mass expolanets are roughly the same size* due to the way Hyrogen compresses. This leads us to the very, very bizarre conclusion that the *smallest gas giant is the heaviest you can get without initiating fusion*. The gas is nowhere near close to escaping it's merely held much closer to itself. *Impressive*.  

So the compression criteria will give us *smallest R* but there's the expansion that takes place which will give us *smallest M*. The method we're working with now will give us smallest M. So let's go for that!

For K=1e5, we found that a central pressure of 1e5 Pascals led to a M/R of 2.821e15 with a radius of 6.226e7 and mass of 1.757e23.

For K=1e6, we find that a central pressure of 1e3 Pascals led to a M/R of 1.892e15 with a radius of 3.669e8 and mass of 6.940e23. 

Despite the central pressures varying wildly, the actual parameters of the mass don't change all that much. 

Anyway, the smallest hydrogen gas giant in terms of radius is going to be whatever one is almost large enough to fuse deuterium, that is a 13 Jupiter mass world. The easiest way to answer this question is just find the "smallest brown dwarf." This is a bit contentious but if we just want a ballpark, abut 0.7 jupiter radius 13 jupiter mass. This is smaller than Saturn. 

And with that... we actually have an answer!
