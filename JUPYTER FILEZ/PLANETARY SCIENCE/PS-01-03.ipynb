This part is actually a mildly interesting question. So, say we start with the part of the surface closest to the sun. # Problem 2.3 \[Back to [top](#toc)\]

*A baseball pitcher can throw a fastball at a speed of ~150 km/h. What is the largest size spherical asteroid of density $\rho = 3 g/cm^3$ from which he can throw the ball fast enough that it...*

*a) Escapes from the asteroid into heliocentric orbit?*

This is as simple as calculating the escape velocity.

$$ v = \sqrt{\frac{2GM}{r}} $$

Now, since we are given the density, we can find r in terms of other quantities. 

$$ M = \rho \frac43 \pi r^3 \rightarrow r = \left( \frac34 \frac{M}{\pi \rho} \right)^{1/3}$$

Seems we have a chocie if we want to quantify the asteroid by its mass or its radius... if we have one we have the other, so both choices are the same. LEt's choose radius since the result is the neatest.

$$ v = \sqrt{\frac{8G\rho \pi r^2}{3}} $$

In truth the problem asks for size, which is r in this case, so we write

$$ r = v\sqrt{\frac{3}{8G\rho \pi}}  $$

We can get an actual number out of this, but we need to be units careful. Put everything in SI.

150 km/h is 0.416 km/s, or 416 m/s

3 $g/cm^3$ is also 3000 $kg/m^3$

G is 6.67e-11 in SI. So our final radius is... 321,299 m, or 321 km. Which is a sizeable rock but not big in astronomical terms, as is to be expected. 



*b) Rise to a height of 50km?*

Now this one we will sove with potential energy! The kinetic energy of the ball is

$$ \frac 12 mv^2 = 86528m $$

Note that we will have to find a way to cancel this mass out later. Fortunately we think we can find that in the potential energy. 

Gravitational potential is given by $-GMm/r$. In this case we will have two different radii, the initial r (the radius of the asteroid itself) and the final r (the point it reaches maximally). The magnitdue of the difference between these two should equal the kinetic energy. So...

$$ 86528 m = \frac{GMm}{R} - \frac{GMm}{r_2} $$
$$ \Rightarrow 86528 = \frac{GM}{R} - \frac{GM}{r_2} $$

Look at that, we have compeletely managed to remove the mass of the ball. What we're actually seeking here is R, $r_2$ is given as 50,000 m. So we can solve... ah, but what we need R, however R is implicit in M, so we have a bit more work to do in actually resolving this. Replace M with our earlier definition. 

$$ \Rightarrow 86528 = G\rho \frac43 \pi R^2 - \frac{G\rho \frac43 \pi R^3}{r_2} $$

This equation will clealry have multiple possible solutions, let's hope only one of them is physical. 

$$ \Rightarrow \frac{86528}{G\rho \frac43 \pi} = R^2 - \frac{R^3}{r_2} $$
$$ \Rightarrow 1.032e11 = R^2(1 - \frac{R}{r_2}) $$

Now keep in mind $r_2$ is actually R + 50,000, since we are considering *from* the surface. This changes things. 

$$ \Rightarrow 1.032e11 = R^2(1 - \frac{R}{R+50000}) $$

It occurs to me that I don't have to solve this algebraically, I *can* solve this numerically. Which is good as we've got a funky denominator situation. (Altenratively we could solve it by approximation, but I like my graphs).

Anyway, the result here is R=2,113,000 m. Which is 2113 km. This is comparable to the radius of Mercury, but we happen to know that Mercury is *very* dense and would have a much stronger gravitational field. (Also at this point the constant density assumption is a bad one). 







*c) Goes into a stable orbit aroudn the asteroid?*

This is a circular motion problem. The force acting on something in circular motion is $\frac{mv^2}{r}$. In this case we might as well be on the surface of the asteroid so r=R, m is going to cancel out quickly, and we know what v is. The Gravitational force will have to be equal to this, $\frac{GMm}{r^2}$. This means we end up with:;

$$ \frac{v^2}{R} = \frac{GM}{R^2} $$
$$ \Rightarrow v^2 = \frac{GM}{R} $$
$$ \Rightarrow v^2 = G \rho \frac43 R^2 \pi $$
$$ \Rightarrow R = v\sqrt{\frac{1}{G\rho \frac43 \pi}} $$

Note that this is very similar to the answer to part a, differing only by a constant factor, $\sqrt{2}$. So we multiply the answer to a) by that and get 454 km, slighlty larger, as it should be.

*The virial theorem states that the time-averaged potential energy of a bound self-gravitating system is equal to two times the negative of the time-average of the kinetic energy of the system, $<E_G> = -2<E_K>$*

*a) State the virial thorem for the N-body problem in words and also in mathematical form.*

Note that the definition applies to *all* self-gravitating systems. So the in words definition still applies. However, mathematically including the N-body situation demands that we add sums over all the energies:

$<\sum E_{GN}> = -2<\sum E_{KN}>$

Add and then take the average, just in case something unusual crops up in the addition. 

*b) Verify the virial theorem for the case of two bodies on a circular orbit.*

For two bodies to be on a circular orbit, we must have a binary situation. Thus, m=M here. The kinetic and gravitational energies of both objects is identical, so we actually only need to show that the theorem holds for one of them. 

Potential is $-GM^2/r$, kinetic energy is $\frac12Mv^2$. Already this seems odd, gravitational potential energy is negative, but the theorem also has a sign difference. The velocity of something in a circular orbit is...

$$ v = \sqrt{\frac{GM}{r}} $$

And so hte kinetic energy is

$$ \frac12 \frac{GM^2}{r} $$

And what do you know, multiply this by negative 2 and you get the gravitational potential energy. Neat! 

No averages were taken as circular orbits do not change their energy.

$$\label{P2.20}$$

*Estimate the amplitude of tides that the Moon raises on Earth. (Hint: integrate the tidal force to compute a tidal potential. Next, calculate the tidal potential of a test particle at the sublunar point on the Earth's surface. Finally, determine the height by which the blob must be raised for the change in its gravitational potential relative to Earth to be equal in magnitude to its tidal potential.)*

Okay, this one seems fun. Since we're estimating, let's use the approximation for tidal force.

$$ F \approx \frac{2xGm}{x_0^3} $$

$x_0$ is the position of our distant gravitationally acting body, x is the point on the body experiencing the force, and m is the distant gravitational pull. We are told to integrate to get a potential. Since we are estimating the moon, the distance of the moon from the earth remains constant, so x is our variable. Thus we get a potential...

$$ U \approx \frac{x^2 Gm}{x_0^3} $$



The problem is worded badly--what we're doing is essentially a potential balancing problem. A particle, to be affected by tides, needs to have the same relative potential in each case. At the surface of the earth, this potential is:

$$ U_G = -\frac{GM_e}{R_e} $$

So without the tides we have the particle resting on the surface of the earth. WITH the tides the potentials add.

$$ U_T = \frac{x^2Gm}{x_0^3} - \frac{GM_e}{x} $$

Naturally for this magnitude to be the same... hold on this implies the magnitude of the earth's gravity needs to get stronger to cancel it out, which is the opposite of what it should be. 

The trick is actually in the *signs*. The actial gravitational force of the Earth here points *downward,* but the potential is opposite in sign to the force in this case, so what we actually need here is a *sign flip* for the gravitational potential. 

$$ U_T = \frac{x^2Gm}{x_0^3} + \frac{GM_e}{x} $$

*now* the Earth's gravity field needs to be weaker to balance it out, the total potential must remain the same. (Why? The problem suggests the same thing, with T = |dG| which implies G - T = newG or something similar, obviously, but the question is why is this the case? It seems that it might just be a consequence of adding potential energy to a system and it seeking to return to its minimum... hmm. Will need to investigate more, tides are important. Also in this first order 1D approximation, the poles don't feel the tidal fordces (and they only feel tiny ones in reality at that) so they would have the same equipotential as before!
