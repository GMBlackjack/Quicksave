# Problem 5 \[Back to [top](#toc)\]
$$\label{P5}$$

*Consider a spherical blackbody of radius R and temperature T. By integrating 9.8 for the radiative flux with $I_\lambda = B_\lambda$ over all outward directions, derive the Stefan-Boltzmann equation in the form of 3.17. You will also have to integrate over all wavelenths and over the surface area of the sphere.*

Okay, let's do this, starting from the top let's *understand* radiative transfer.

Let's be clear here, I is the intensity, a quantity that is the same no matter how far away we are, for it is a per solid angle value. B is the blackbody radiation. 

Our goal is 3.17 $L = 4\pi R^2 \sigma T_e^4$, which is one way to define effective temperature. L is luminosity, $\sigma$ is the Stefan-Boltzmann constant. Near this equation is $\mathcal F_{surf} = \sigma T_e^4$, but we'll see if this is actually relevant.

Anyway, we are given a suggestion of where to start: 9.8.

$$ \mathcal F_\lambda d\lambda = d\lambda \int I_\lambda cos\theta d\Omega = d\lambda \int I_\lambda cos\theta sin\theta d\theta d\phi  $$

Now, I=B, so what we really need is B. We can find this from 3.22.

$$ B_\lambda = \frac{2hc^2}{\lambda^5} \frac{1}{e^{hc/\lambda kT}-1} $$

There is no angular dependence here so we can just plug it in and be done. However, the angular dependence pretty clearly goes to zero due to the nature of the steradian. So, instead, let's integrate with respect to the wavelength first.

$$ \mathcal F = \int \frac{2hc^2}{\lambda^5} \frac{1}{e^{hc/\lambda kT}-1} cos\theta sin\theta d\theta d\phi d\lambda  $$

Actually manually doing this integral is a pain. Fortunately we have computers. Ah, as perhaps should be expected, this is not a simple one.



How unsightly. Forunately we are integrating over very specific bounds, 0 to infinity, as wavelength cannot be negative. So if we can identify every part we can just evaluate. This means we need to know what Li is... the logarithmic integral. What's important are the properties. The value at 0 also goes to 0, but the infinite grows without bound. 

Most everything either cancels or is canceled as we approach the limits, the ones that don't become nothing at 0 and infinity are... ...divergent UGH this is annoying but there's a reason we came back here after all. Perhaps try relating B to L next time? Oh my goodness, there *is* an actual integration to be done.

http://people.tamu.edu/~kevinkrisciunas/planck.pdf

But it most certainly is *not* an *easy* one. Let's see if we can figure it out from thsi reference. 



FIRST of all, we can do the solid angle integral, since we actually are only dealing with *half of the flux area*. We could have deduced this, for we were being asked for a relation that gives us L, that is the Luminosity, which is an *observed* quantity. Only half of a spherical blackbody will actually emit that. The angular integral becomes $\frac 12 sin^2\theta$

which when evaluated at $\pi/2$ gives us 1/2. Multiply by the azimuthal angle's integral and we get $\pi$. 

$$ \mathcal F d\lambda = \pi B_\lambda d\lambda $$

And the next step depends on integrating B over all wavelenghts (0 to $\infty$). Usually this is just accepted as $\sigma T^4$ but this is our *goal*. We need to show that this is the case. One of the most helpful things in the document is that $\sigma$ can be made out of other constants. Let's pull some out by clever u-substitution.

$$ \mathcal F = \pi \int_0^\infty \frac{2hc^2}{\lambda^5} \frac{1}{e^{hc/\lambda kT}-1}  d\lambda $$

This is our integral. Let us declare a new variable, one that is equal to the exponent in e, call it x, naturally.

$$ x = \frac{hc}{\lambda kT} $$
$$ dx = -\frac{hc}{\lambda^2 k T} d\lambda $$

Note that this substitution does change the order of integration.

$$ \Rightarrow \mathcal F = \pi \int_\infty^0 -\frac{2hc^2}{\lambda^5} \frac{kT\lambda^2}{hc} \frac{1}{e^x-1} dx $$

$$ \Rightarrow \mathcal F = \pi \int_0^\infty \frac{2ckT}{\lambda^3} \frac{1}{e^x-1} dx $$

now see the annoying part here is that we haven't completley gotten rid of $\lambda$ as it was contained in itself, but we *can* actually do it, for we have the replacement for x! 

$$ \Rightarrow \mathcal F = \pi \int_0^\infty 2ckTx^3 \frac{k^3T^3}{h^3c^3} \frac{1}{e^x-1} dx $$

$$ \Rightarrow \mathcal F = \pi \frac{2k^4}{h^3c^2} T^4 \int_0^\infty  x^3\frac{1}{e^x-1} dx $$

We have the desired $T^4$ dependence. Also we can get $\sigma$ via $\sigma = \frac{2\pi^5k^4}{15c^2h^3}$. 

$$ \Rightarrow \mathcal F = \frac{15}{\pi^4} \sigma T^4 \int_0^\infty x^3\frac{1}{e^x-1} dx $$

An attempt was made to integrate this on my own but HAH no turns out, might not even have an indefinite form, what fun. So we return to our reference. Holy cow, we need the Gamma Function, Zeta Function, *and* Fourier Series for this (wait, maybe not? Perhaps just recognizing the series pattern is sufficient). What kind of insanity...?

So, try to follow this mess.

A known integral is

$$ \int_0^{\infty} e^{-bx}x^{s-1} dx = \frac{\Gamma(s)}{b^s} $$

It's a mess but consider it known. So if we let5 b = a+n

$$ \int_0^{\infty} e^{-(a+n)x}x^{s-1} dx = \frac{\Gamma(s)}{(a+n)^s} $$

Now to try to force the Zeta functiuon we perform infinite sums on both sides with respect to n. Yes, the infinite sum is an operator we can just use! Who'd'a'thunk?

$$ \sum_n \int_0^{\infty} e^{-(a+n)x}x^{s-1} dx = \sum_n \frac{\Gamma(s)}{(a+n)^s} $$

Anyway the right hand side is basiclaly the definition of the Zeta function times the Gamma funciton.

$$ \sum_n \int_0^{\infty} e^{-(a+n)x}x^{s-1} dx = \zeta(s,a) \Gamma(s) $$

Move the remaining sum into the integral, it only acts on a small part.

$$ \int_0^{\infty} x^{s-1} e^{-ax} \sum_n e^{-nx} dx = \zeta(s,a) \Gamma(s) $$

That infinite sum is of the form $1 + q + q^2 ... = \frac{1}{1-q}$. 

$$ \int_0^{\infty} x^{s-1} e^{-ax} \frac{1}{1-e^-x} dx = \zeta(s,a) \Gamma(s) $$
$$ \int_0^{\infty} x^{s-1} e^{-ax} \frac{1}{e^x-1} dx = \zeta(s,a) \Gamma(s) $$

Now this is workable! a=0, s=4

$$ \int_0^{\infty} x^{s-1} e^{-ax} \frac{1}{e^x-1} dx = \zeta(4,0) \Gamma(4) $$
$$ = 6 \sum_n \frac{1}{n^4} $$

Now we could evaluate this. Or we could look up the result of that series since this is not a math course this is a physics course. $\pi^4/90$. So the full answer is...

$$ \int_0^{\infty} x^{s-1} e^{-ax} \frac{1}{e^x-1} dx = \frac{\pi^4}{15} $$

# Problem 8 \[Back to [top](#toc)\]
$$\label{P8}$$

*In Example 9.2.3 suppose that only two measurements of the specific intensity, $I_1$ and $I_2$ are available, made at angles $\theta_1$ and $\theta_2$. Determine expressions for the intensity $I_{\lambda,0}$ of the light above Earth's atmosphere and for the vertical optical depth of the atmosphere $\tau_{\lambda,0}$ in terms of these two measurements.*

To fully grasp this quesiton, we need to consider what the example is actually about. The example just declares the following equation:

$$ I_\lambda = I_{\lambda,0} e^{-\tau_{\lambda,0} sec\theta} $$

Well, the idea is to turn this into tqo equations, two unknowns, with our observed values.

$$ I_1 = I_{\lambda,0} e^{-\tau_{\lambda,0} sec\theta_1} $$
$$ I_2 = I_{\lambda,0} e^{-\tau_{\lambda,0} sec\theta_2} $$

Now we have two equations and two unknowns. Since it's clear what they are, let's leave off the subscripts.

$$ I_1 = I e^{-\tau sec\theta_1} $$
$$ I_2 = I e^{-\tau sec\theta_2} $$

The problem is... this is not a simple "subtract and solve!" situation, the exponential is the exponential. There's not even a sum to exploit to simplify the exponent! However, one thing we *can* do is take the natural logs of both sides... after we move I to the other side.

$$ I_1/I = e^{-\tau sec\theta_1} $$
$$ I_2/I = e^{-\tau sec\theta_2} $$

and ln...

$$ ln(I_1) - ln(I) = -\tau sec\theta_1 $$
$$ ln(I_2) - ln(I) = -\tau sec\theta_2 $$

Now we can combine in our usual, expected way. Canceling out ln(I) is going to be pathetically easy.

$$ ln(I_1) - ln(I_2) = -\tau sec\theta_1 + \tau sec\theta_2  $$
$$ \Rightarrow \frac{ln(I_1) - ln(I_2)}{sec\theta_2 - sec\theta_1} = \tau $$



Now we find the solution for I.

$$ [ln(I_1) - ln(1)]sec\theta_2 - [ln(I_2) - ln(I)]sin\theta_1 = 0 $$
$$ \Rightarrow ln(I)(sec\theta_1 - sec\theta_2) + ln(I_1)sec\theta_2 - ln(I_2)sin\theta_1 = 0 $$
$$ \Rightarrow ln(I) = \frac{ln(I_2)sec\theta_1 - ln(I_1)sin\theta_2}{sec\theta_1 - sec\theta_2} $$
$$ \Rightarrow I = e^{\frac{ln(I_2)sec\theta_1 - ln(I_1)sin\theta_2}{sec\theta_1 - sec\theta_2}} $$

Which is kind of ugly but it is the answer.

THe optical thickness is actually something easy to answer. When viewed dead-on the thickness must be less than 2/3 to be able to see the star at all, put presumably must rise to above 2/3 to be significant at the limb.So, in essence, dead-on the gas is mostly clear, but when viewed at an agnle the distance traveled is sufficient to stand out. 

The question is, what physical situation produces such a vibrant effect? In essence, visually the center of the star is bright, then it gets dark, then we get bright agian to see the ring. Well, this can happen if the atmosphere is highly emissive, but no *so* emissive that it drowns out the star's light when it is in front of it. Granted making everything work just right for this situation is a bit odd. In other news this system would have to be highly specific. 

That said we do observe this sort of thing with planetary atmospheres sometimes depending on context, though this is often because light is being taken from the sun through the atmosphere, and isn't really spherically symmetric. # Problem 15 \[Back to [top](#toc)\]
$$\label{P15}$$

*Derive 9.35*

$$ I = I_0 e^{-\kappa \rho s} + S(1-e^{-\kappa \rho s}) $$

Naturally basically everything is wavelength dependent.  

The first part makes sense, as it is the slow attenuation of the initial intensity based on the opacity and the current density. But the Source Function? Well, this is mentioned as a specific solution of the transfer equation 9.34. So let's look at that.

$$ -\frac{1}{\kappa \rho} \frac{dI}{ds} = I - S $$

s is, of course, the length. In this form this is a differential equation, and one that's not that bad either. If we assume that S doesn't depend on the distance, which is reasonable if we're going through a single medium. (Medium changes would mess this up pretty quick). So if everything's a constant, what we essentially have is a differential equation of the form::

$$ \frac{dI}{ds} = A(I-S) $$

Where A just holds constants. Taking the integral is not helpful here, so perhaps we can go for separation of variables.

$$ dI = A(I-S)ds $$
$$ \Rightarrow \frac{1}{I-S}dI = Ads $$
$$ \Rightarrow ln|I - S| = As+C $$
$$ \Rightarrow |I - S| = e^{As+C} $$
$$ \Rightarrow |I - S| = e^{As}e^C $$
$$ \Rightarrow |I - S| = Ce^{As} $$

That absolute value sign is going to be quite annoying. Let's just write down both versions. 

$$ I = Ce^{As} \pm S $$

We can replace A with what we know it to be.

$$ I = Ce^{-\kappa \rho s} \pm S $$

Now, C is unknown. We need to find some kind of boundary condition we can set it to. One thing we do know is that at s=0 the result is $I_0$. So...

$$ I_0 = C \pm S \Rightarrow C = I_0 \mp S$$

And so our full solution...

$$ I = (I_0 \mp S)e^{-\kappa \rho s} \pm S $$
$$ \Rightarrow I = I_0e^{-\kappa \rho s} \pm S(1 - e^{-\kappa \rho s}) $$


So, now there is just one thing left we have to justify: why resolve $\pm$ to +? The condition for each relies on the magnitude of I with respect to S. But, the thing is, at any given time S cannot be larger than I! Now, we know this to be physically true by argument, because of course inensity approaches the source function, it has to. But is that arguing based on knowing the result ahead of time, and thus a fallacy? Let's investigate a little closer.

Well, let's try to construct a situation where the intensity is less than the source function, we shouldn't be able to. The obvious case is "what if there is no incoming radiation and $I_0$ is zero? Then obviously I starts at zero, enters the medium, and has to increase over time." Yes, but that consists of a *medium change* which is not quantified here as $\kappa$ is constant. What if we are already in the medium when we start? Then the backing is *already S!*. In fact, if we are always within a medium, thus making $I_0 = S$, then we get a cancelation, if this cancelation was using the minus sign, I = -S, which is physically nonsense! Thus, we must use "+".

# Problem 16 \[Back to [top](#toc)\]
$$\label{P16}$$

*The transfer equation 9.34 is written in terms of the distance, s, measured along the path of a light ray. In different coordinate systems the transfer equation will look sliightly different, and care must be taken to include all of the necessary terms.*

*a) Show that in a spherical coordinate system, with the center of the star at the origin, the transfer equaiton has the form*

$$ -\frac{cos\theta'}{\kappa \rho} \frac{dI}{dr} = I-S $$

*where $\theta'$ is the angle between the ray and the outward radial direction. Note that you cannot simply replace s with r!*

This is actually not all that difficult if we've already considered the differences between the two. There is a question as to what the angle means... angle away from the radially outward direction? Every angle is radially outward! Unless, of course, we're considering the angle with respect to the *observation* point. In which case this is simpler, but doesn't seem to produce the above result... hmm. 

New thought: the angle changes as we move (which it really should), and the angle reported above is the angle that describes hwo perpendicular the motion is relative to the center of the star. This way we can easily extend the line behind the star and anywhere, really. This is decidedly not *normal* spherical coordinates, which would have an angle of 0 where this has an angle of 90, and the angle changes based on position. 

In fact this is odd enough I think we need a picture.



So at closest approach, an infinitesimla change in r does nothing to the actual length of the line, but at far distances the line is growing almost exactly with r. The angle is 90 degrees at zero, but approaches 0 and 180 at the maximum. Which is to say, the actual vector of change *along* the line goes like cosine. Which means we don't evne have to break out any math, this is just how vector quntities relate to each other. Bit of an odd choice, admittedly. 

I will still point out that this is NOT a spherical coordinate system, this is a coordinate sytsem with $\theta'$ instead of proper $\theta$. If we uased actual theta, the perpendicular portion would be at angle zero, and the "Far" distances at plus or minus 90. So if we used the *coordinate* angle, the result would be related to sine.

$$ -\frac{sin\theta}{\kappa \rho} \frac{dI}{dr} = I-S $$

There are no sign flips as our "domain" has just shifted from 0-180 to -90-90. I suppose the use of negative angles might make it easier to use the $\theta'$ definition in some cases, or it might just depend on preference. 

*b) Use this form of the trnasfer equation to derive 9.31*

$$ \frac{dP_{rad}}{dr} = -\frac{\bar \kappa \rho}{c} F_{rad} $$

This relates the pressure to the energy. Let's see how this goes by just following the book's procedure for finding it. First of all, it assumed plane-parallell gray atmosphere, so we shall do the same. ANyway, first, rewrite everything in terms of optical depth. Ah, but we're in a different coordinate system now! So...

$$ d\tau = -\kappa \rho ds $$

Is not our relation anymore. We have a dr. But we do kind of already have a relation between ds and dr, becasue obviously if we transfer back we have to reclaim the original, so $ds = \frac{1}{cos\theta'} dr$. Which means the conversion to optical depth is the same as normal.

$$ \frac{dI}{d\tau} = I-S $$

So 9.36 still holds. However, we need a *vertical* optical depth, $\tau_\nu$. THe relation between them remains the same if we assume plane atmospheres, which at relatively small scales (such as for planets or stars) holds.

$$ \tau = sec\theta \tau_\nu $$

Keep in mind this angle is entirely independent of the angle in the first part to this problem. Anyway, the next step also holds, as the conversion to vertical optical depth is still maintained. 

$$ cos\theta \frac{dI}{d\tau_\nu} = I-S $$

At this point we supposedly integrate all quantities over the entire wavelength but we already have the subscripts dropped so this doesn't change anything. *inside* of $\tau_\nu$ we replace $\kappa$ with $\bar\kappa$. This is what the "gray atmosphere" assumption allows. Integrating both sides over solid angles or with trig functions and solid angles is reminiscent of problems done in the PS-01 notebook. This gets us to:

$$ \frac{dP_{rad}}{d\tau_\nu} = \frac{1}{c} F_{rad} $$ 

At first it seems like we've failed here, we're at the same form as before! But, aha, we just need to state it in spherical coordiantes, which means *expanding out* what we already have. Simple! But let's be thorough about it.

$$ d\tau_nu = \frac{1}{sec\theta} d\tau = -\frac{\kappa\rho}{sec\theta} ds =  -\frac{\bar\kappa\rho}{sec\theta cos\theta'} dr $$.

Now, the one issue here is that, theoretically, we have already integrated over the angles. However, something strange is that secx(cosx) is just... 1, since sec is inverse cosine. But we have $\theta$ in one and $\theta'$ in the other, how do we justify this? Well, the angle $\theta$ is not the coordinate angle either, it is the angle measured from the vertical optical depth to the actual observed optical depth. That "vertical line" is exactly the radial direction, so we end up with the exact same coordinate in both! Thus, it cancels to one, and we have...

$$ -\frac{1}{\bar\kappa \rho} \frac{dP_{rad}}{dr} = \frac{1}{c} F_{rad} $$ 
$$ \Rightarrow \frac{dP_{rad}}{dr} = -\frac{\bar\kappa \rho}{c} F_{rad} $$ 

Which is precisely what was sought.

# Problem 17 \[Back to [top](#toc)\]
$$\label{P17}$$

*For a plane-parallel atmosphere, show that the Eddington approximation leads to expressions for the mean intensity, radiative flux, and radiation pressure given by 9.46 - 9.48.*

The Eddington Approximation is a bit vague, so we did some reasearch on it. All that's stated here is there's $I_{in}$ and $I_{out}$ and that they differ at every level, and at the top of the atmosphere $I_in$ is zero. The image (fig 9.15) confused us initially, but it is simply showing that there is more energy being radiated "outward" form a point than "inward." In fact the inherent assumption is that energy *flows outward*. 

So, now that we quantified specifically waht the approximation is, what is the first thing we have to show?

$$ <I> = \frac12(I_{out} + I_{in}) $$



So, what exactly the "mean intensity" is, we choose to think of it like equation 9.3.

$$ <I> = \frac{1}{4\pi} \int I d\Omega $$

At the top half of the sphere we have the outgoing intensity, and on the bottom half we have the inward intensity. So this integral will split up in our case.

$$ \Rightarrow <I> = \frac{1}{4\pi} \int I d\Omega = \frac{1}{4\pi} \int I sin\theta d\theta d\phi $$
$$ = \frac{1}{4\pi} \left[ \int_0^{2\pi}\int_0^{\pi/2}I_{out} sin\theta d\theta d\phi + \int_0^{2\pi}\int_{\pi/2}^{\pi} I_{in} sin\theta d\theta d\phi \right] $$
$$ = \frac{1}{2} \left[ \int_0^{\pi/2}I_{out} sin\theta d\theta  + \int_{\pi/2}^{\pi} I_{in} sin\theta d\theta \right] $$
$$ = \frac{1}{2} \left[ I_{out} + I_{in} \right] $$

This is what you'd *expect* the average intensity to be, the sum of the two contributions divided by two, so yay!

Next?

$$ F_{rad} = \pi (I_{out} - I_{in}) $$

So the radiative flux is based not only on a difference, but a differene times $\pi$ of all things. But, see, for the flux we don't need to divide by the solid angle after we're done, for it's a *per solid angle* quantity. So when we take the integrals we did above, the $\pi$ will not cancel out.

The full integarl is

$$ F = \int\int I cos\theta sin\theta d\theta d\phi = 2\pi \int I cos\theta sin\theta d\theta $$

Now sincos is not going to be positive over the whole domain, the other half will be negative, and furthermore tha amplitude of the function is 1/2 rather than 1. I say it this way so I can avoid writing out the integral when it's really a basic thing that can be reasoned out. 

$$ F = \pi (I_{out} - I_{in}) $$

Where the minus sign comes from the fact that over the domain $\pi/2 -\pi$ the integral becomes negative. Under the curve, after all. 

So, lastly, we have:

$$ P_{rad} = \frac{2\pi}{3c}(I_{out}+I_{in}) = \frac{4\pi}{3c}<I> $$

Where that escond step just invovles using the definition of the first thing we proved. One might think we need to relate F to P, but I don't think so. Let's go back and look up some more basic determinators for P, I think I remember them existing. 

We found it in 9.9

$$ P = \frac{1}{c} \int I cos^2\theta sin\theta d\Omega $$

Which is going to be positive over the entire integral and have equal areas on both domains, so naturally we'll end up with additions. The integral goes to $\frac{-1}{3} cos^3\theta$ with the $2\pi$ from azimuthal, getting us

$$ P = \frac{2\pi}{3c}(I_{out} + I_{in}) $$

And that's all we need to find. Integrals, everyone!

# Problem 18 \[Back to [top](#toc)\]
$$\label{P18}$$

*Using the Eddington Approximation for a plane-parallel atmosphere, determine the values of $I_{in}$ and $I_{out}$ as functions of the vertical optical depth. At what depth is the radiation isotropic to within 1%?*

Oh! Well this is a bit unexpected. But we think we can work backward.

First of all, we have the optical depth related to the average intensity as:

$$ \frac{4\pi}{3} <I> = F (\tau_\nu + \frac23) $$

Now we can extract things from both F and the average I here using the relations from the previous problem.

$$ \frac{2\pi}{3} (I_{out}+I_{in})  = \pi (I_{out} - I_{in}) (\tau_\nu + \frac23) $$
$$ \Rightarrow \frac{2}{3} (I_{out}+I_{in})  = (I_{out} - I_{in}) (\tau_\nu + \frac23) $$

This can actually be simplificed considerably.

$$ \Rightarrow \frac{4}{3} I_{in}  = (I_{out} - I_{in}) \tau_\nu $$

If we fiddle around we can actually find a ratio relation.

$$ (\frac{4}{3}+ \tau_\nu ) I_{in} = \tau_\nu I_{out} $$
$$ \Rightarrow (\frac{4}{3\tau_\nu}+ 1) I_{in} = I_{out} $$
$$ \Rightarrow (\frac{4}{3\tau_\nu}+ 1) = \frac{I_{out}}{I_{in}} $$

Hold the phone, we can answer the vertical optical depth question here. 

$$ \Rightarrow (\frac{4}{3\tau_\nu}+ 1) = \frac{I_{out}}{I_{in}} $$

Naturally to make it isotropic the fractional term needs to be as small as possible, which means $\tau_\nu$ needs to be large. We specificlaly seek a 1% difference, so...

$$ \Rightarrow (\frac{4}{3\tau_\nu}+ 1) = 1.01 $$
$$ \Rightarrow \frac{4}{3\tau_\nu}= 1/100 $$
$$ \Rightarrow \frac{3\tau_\nu}{4}= 100 $$
$$ \Rightarrow \tau_\nu = 400/3 $$

Which is to say *extremely deep*, 2/3 is what you *see*. That said in a star this is not unreasnable to find, given how much the density increases as you go down. Then again the Eddington Approximation might break down at this point. 

So, let's assume we know the equilibrium temperature, we think we can use this to find another relation. 

$$ <I> = \frac{3\sigma}{4\pi} T_e^4 (\tau_\nu + \frac23)$$
$$ F = \sigma T^4_e $$

And so...

$$ \frac12 (I_{out} + I_{in}) = \frac{3\sigma}{4\pi} T_e^4 (\tau_\nu + \frac23)$$
$$ \pi (I_{out} - I_{in}) = \sigma T^4_e $$

Now one might think to do a substitution, but this just gets us back to where we alreayd were. So, we solve these as they are. 

$$ I_{out} = \frac{3\sigma}{2\pi} T_e^4 (\tau_\nu + \frac23) - I_{in}$$
$$ I_{out} = \frac{\sigma T^4_e}{\pi} + I_{in} $$

And so...

$$ \frac{3\sigma}{2\pi} T_e^4 (\tau_\nu + \frac23) - I_{in} = \frac{\sigma T^4_e}{\pi} + I_{in} $$
$$ \Rightarrow \frac{3\sigma}{2\pi} T_e^4 (\tau_\nu + \frac23) - \frac{\sigma T^4_e}{\pi} = 2I_{in} $$
$$ \Rightarrow \frac{3\sigma}{4\pi} T_e^4 (\tau_\nu) = I_{in} $$

Hey, that's 0 when the vertical optical depth is zero, as it should be!

Finding the other one is somewaht trivial, as we have a very simple relation for it.

$$ I_{out} = \frac{3\sigma}{4\pi} T_e^4 (\tau_\nu + \frac43) $$

*Consider a horizontal plane-parallel slab of gas of thickness L that is maintained at a constant temperature T. Assume that the gas has optical depth $\tau_{\lambda,0}$, with $\tau_\lambda = 0$ at the top surface of the slab. Assume further that no radiation enters the gas from outside. Use the general solution of the transfer equation 9.54 to show that when looking at hte slab from above, you see blackbody radiation if $\tau_{\lambda,0} >> 1$ and emission lines (where $j_\lambda$ is large) if $\tau_{\lambda,0} << 1$. You may assume the source function, $S_\lambda$, does not vary with position inside the gas. You may also assume thermodynamic equilibrium when $\tau_{\lambda,0}$ >> 1*

Let's grab 9.54 and take a look.

$$ I_{\lambda}(0) = I_{\lambda,0} e^{-\tau_{\lambda,0}} - \int_{\tau_\lambda,0}^{0} S_\lambda e^{-\tau_\lambda} d\tau_\lambda $$

So, if the source function does not vary with position inside the gas, it's a constant, and thereby can be pulled out of the integral.

$$ I_{\lambda}(0) = I_{\lambda,0} e^{-\tau_{\lambda,0}} - S_\lambda  \int_{\tau_\lambda,0}^{0} e^{-\tau_\lambda} d\tau_\lambda $$

In the large optical depth case, the exponentials go to zero, leaving only the "0" evaluation on the integral bounds. Which, as the exponent is negatrive, will flip the sign. THus...

$$ I_{\lambda}(0) = S_\lambda $$

And as we know in LTE S=B, so this is blackbody radiation. 

But what about the *other* case? What if we make it arbitrarily small?

Well, in that case, the integral vanishes, there's not a dirac delta hiding in the exponential. So we end up with...

$$ I_{\lambda}(0) = I_{\lambda,0}$$

Which is to say we only get the background radiation. But wait, how does this translate to seeing emission lines? Well, as things are optically thin, that means any light produced feels as though the medium is effectivley transparent. Thus, anything emitted *by* the medium will be seen everywhere within the medium, making the initial I the same everywhere. This would be true even if there were no external source of radiation, only internal.

Note that S is a source function, and that it is driven largely by blackbody interactions, so it has to *absorb* radiation to do anything. If it can't absorb anything, nothing happens. 

That said, we do note that we haven't factored j in at all here, so perhaps we were too hasty to say the integral went to zero. Let's do a better order approximation.

$$ I_{\lambda}(0) = I_{\lambda,0}e^{-\tau_{lambda,0}} - S_\lambda (-1 + e^{-\tau_{\lambda,0}})$$

Now, to first order, this subtrracts to nothing. First order gives us $1 - \tau_{\lambda,0}$ which would in turn reduce this to...

$$ I_{\lambda}(0) = I_{\lambda,0}(1 - \tau_{\lambda_0}) + S_\lambda (\tau_{\lambda,0})$$

So naturally $\tau$ here is small and so those terms would vanish to first approximation. To second approxiamtion, though, we *can* influence things by adjusting the value of S. Specifically, S=$j/\kappa$, and j is the emission coefficient. AT wavelengths where j is large, this term at the end will not be neglegible, even at small optical depths. So *this* is where you can get emission lines. Granted, S has to be large for this to occur, otherwise its insignificant. 

We realize later that we were supposed to assume $I_{\lambda,0} = 0$. Oops. Everything still holds though.
