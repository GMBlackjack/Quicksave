Time to actually try to understand the Mean Free Path.

$$ l = \frac{1}{\sigma n} $$

n is the number of particles in a unit volume, which in our case is 5 protons per cubic centimeter, or 10 particles as there are also free electrons. $\sigma$ is the average cross sectional area. The cross sectional area is dominated by protons, so half of that. Proton diameter is 0.8e-15 m ish, which is 0.8e-13 cm. Which gives an area of 2e-26 square centimeters roughly, and we divide by two to account for the half-electrons. Thus we get a mean free path of... 9.9e24 cm. That's 9.9e19 km. 

The earth's radius is about 6e3 km, and the distance of the bowshock is about 20 times this, so 1.2e5 km. This is peanuts to this. Space is basically EMPTY, friend, ain't getting hardly any collisions. (The magnetosphere is inside the bow shock).

*Calculate the speed at which meteoroids with the following heliocentric orbits will encounter the Earth's atmosphere.*

*a) An orbit very similar to that of the Earth, so that $v_{inf} << v_{e}$ and thus $v_{impact} \approx v_e$*

So essentially we can model this as the Earth dragging an object in from infinity. 

The magnitude of gravitational potential is GMm/r and we correlate this with kinetic energy of $mv^2/2$ to get 

$$ \sqrt{\frac{2GM}{r}} = v $$

In this case we only need r, the height of the atmosphere. Earth's radius is 6,371km. As for the size of the atmosphere, well, it's between 100 and 10,000 km. At about 100km the pressure is only 1e-2 Pa, so let's consider this where it ends. So basically it's a tiny shell on the Earth, giving us a total radius of 6,471 km. So by the time this meteor hits the atmosphere it's going... 11km/s. This is roughly the minimum speed of a meteor.

Now this is for something not really moving relative to Earth at the start, the earth itself moves about 30km/s, so that'll change things for other situations. 

*b) A parabolic orbit with perihelion of 1 AU and i = 180*

Okay so the perihelion is the point at which the object is moving the fastest. This is a parabolic trajectory so the eliptical orbit math no longer qualifies. A quick look up reveals that a the speed of a parabolic trajectory can be given by:

$$ v = \sqrt{2} v_0 $$

Where $v_0$ is the speed of an object in a circular orbit. And what do you know, the Earth is in a mostly circular orbit, so it's 30km/s! So 42km/s for our buddy here.

HOWEVER, what *direction* are we coming in at? i is the inclination. If it were 0, it would be in line with the Earth. As it is 180 degrees, we are going for a *head on collision* oh boy! Simple adding of hte velocities provides 72 km/s. This is how fast the Earth sees the meteor coming in from a distance. However, does Earth's gravitational field add anything? GM/r gives the change in potential energy no matter what, which is 6.159e7 every time. (ignoring the small mass m contribution). This correlates to the change in $v^2/2$. So the part that adds is not v+v, but $v^2+v^2$. We know that the "capturing" speed is 11km/s, so once we add the square contributions the new speed is... 73km/s. Barely even an increase.  

*c) A parabolic orbit with perihelion of 1 AU and i = 0*

Okay so now we're going the same direction as Earth, so the observed speed is 42-30 = 12. This one will add significantly due to Earth's gravitational field, giving 16km/s. 

*d) A parabolic orbit with perihelion of 1 AU and i = 90*

Okay so now we're coming in directly perpendicular to Earth's motion, so Earth's motion is largely irrelevant. (There's probably some higher order term that arises from hitting it at speed and the divergence of the meteor's path, but let's ignore that). Thus the Earth sees just a plain 42km/s object heading its way. This increases only to 43km/s due to the potential energy shift. 

*e) An orbit with a=2.5 AU, e=0.6, and i=0*

Okay so this one is eliptical. So, let's look to see if any of the obvious candidates are actually at 1 AU. 

a is 2.5 AU.

ea is the distance from the ellipse center to the sun, so 1.5 AU.

Thus the distance at perihelion is 1 AU. Hah! So closest approach is once again what we're looking for. So we can actually use the known formula:

$$ v = \sqrt{\frac{GM}{a} \frac{1+e}{1-e}} $$

So the actual speed we have is... 38 km/s. 

At i=0 we are incoming *slowly* and thus end up with an apparent speed of 8, and an ending speed of 14km/s.

*f) An orbit with a=2.5AU, e=0.6, and i=30*

Hey, this ain't in line with anything! Well, fortunately, we can scale the contributions of the speeds like vectors. The new square speed can be written as:

$$ v^2 = (v_m-v_ecosi)^2 + v_a $$

$v_m$ is meteor speed relative to sun, $v_e$ is earth speed relative to sun, and $v_a$ is the speed we have just from being dragged through Earth's gravitational field. Note the negative. When i=0 the speeds interfere "destructively", while at i=180 it's "constructive". Anyway at i=30 with a meteorite speed of 38 km/s we get an observed speed of 12km/s which means an ending speed of 16km/s, similar to our answer for c), we believe by coincidence.

# Problem 8.3 \[Back to [top](#toc)\]
$$\label{P8.3}$$

*a) Calculate the pressure on a meteor moving at a speed of 10 km/s at an altitude of 100km above the Earth's surface.*

So, atmospheric pressure is a useless metric, as whatever pressure it provides is peanuts to the pressure felt by this ballistic monstrosity. What does matter is the atmospheric *density*, that is, how much substance is there in our way?

Force is $\frac{kg m}{s^2}$, as a reminder, and pressure is the force over area, so curiously $kg/ms^2$, so keep this in mind for our unit analysis. In theory we can find density at altitude, that is, $kg/m^3$. Call this $\rho$. We will need this later.

For now, consider just a single particle hitting the meteor. The speed at which it hits might as well be the speed of the meteor itself since ambient atmosphere temperatures are going to be low, so it'll hit with precisely that speed and (presumably) bounce right off. This imparts a momentum change of 2mv per particle, where m is the mass of one particle. Recall that force is the time derivative of momentum, so this change in momentum is precisely what we need for one particle. However, we have a lot more than one particle striking every second. How many is it every second? Well that would be $\rho^* v$, the density times the velocity, which gives us a resulting unit $1/m^2s$. Which is basiclaly "number of particles hit per unit area in a second." 

So, $2m\rho^*v^2$ is our result. Note the * on $\rho$, as technically we found above the *number* density. The mass density would be this times the mass per particle, so here we go.

$$ 2\rho v^2 $$

Now, though, technically speaking this is only for particles directly incident on the meteor. Fortunatley since we're going so fast, particles coming in at an angle provide almost nothing, and thus we can simply declare this to be the pressure. 

Research suggests the 2 isn't there. Why? Others imply that it should just be 1/2 instead. Oh bother...

Okay, so, in the case of a cube, the pressure has a 1/3 in front of that expression. One can imagine why: each face feels equal pressure but each face has to get pressure from different components in the gas, hence a 1/6 deal here, even in the case of perfectly elastic collisions. In a 2D environment, we'd have 4 different faces, so 1/2. In a "1D" environment, the 2 different faces would remove the initial "2." 

Now, in *our* case, there is only *one* face, and it is traveling much faster than the molecules. So from the meteor's perspective all molecules are moving *one* direction, so *all* of the energy is focused on the meteor. Thus, we keep the 2. 

Anyway now we need to find the air density. We could use the barometric formula for this. Or we could just look it up at 100km. 5.77e-7 $kg/m^3$ it turns out. 

Anyway this ends up giving us 115 Pa. 

*b) Repeat for a meteor at the same speed 10km above the Earth's surface.*

Density here is 4.127e-1, so we get 8.26e7 Pa. Which is to say, a lot. 

*c) Repeat your calculations for a) and b) for a meteor traveling 30km/s*

This change makes the upper atmosphere pressure 1040 Pa.

The lower atmosphere pressure is 7.434e8 Pa. 

For the record normal atmospheric pressure at those elevations are 1e-1 Pa and 1e4 Pa, higher to lower. In every case the meteor's motion is eclipsing the natural motion of the particles by orders of magnitude. 

*d) The tensile strength of comets are of order 1e4 dyne/$cm^2$, the strengths of chondrites are roughly 1e6, stronger stony objects have strengths approximately 1e9, whereas iron impactors have effective strengths of about 1e10. Compare these tensile strenghts to the pressures calculated in parts a) - c) and comment.*

So, first of all, why use this freaky unit? Each one is 1e-5 pascals, so to adjust:

COMET: 1e-1 Pa

CHRONDITE: 1e1 Pa

STONY: 1e4 Pa

IRON: 1e5 Pa

So, comets and chrondites start to break up almost instantly, both are below the 115 pascals for our lowest situation. Stonys and irons will survive at high altitudes, but eventually break up as they get closer and closer to the ground, and at 10km we expect explosions from basically everything. Except at that height the speed of the object is still measured in km/s so it'll explode then hit quite well.

One thing that's not being noted here is that the atmosphere is going to slow things down. The pressure that isn't enough to tear the rock to shreds is going to lower its speed, and thus lower the stress on it at lower elevations. Earth is moving at roughly 30km/s so the average meteor will strike the atmosphere at that speed, gaining or losing some speed due to gravity, and then dealing with the pressures of the air on it. 

A quesiton one might ask is at what pressure does the meteor start igniting the air aroudn it--likely a very different question. 

ADDENDUM: oh hey we found the equation we were supposed to use.

$$ P = \frac{C_D \rho v^2}{2} $$

Where we note the difference is the drag coefficient. What is the drag coefficient of a sphere? About 0.47. So we have significantly overestimated our pressures. One may think "but we did it from first principles!" Yes, but air is a fluid. Drag coefficietns are genreally experimentally determined, also they change with Reynold's numbers as well. Not even a "flat" wall is good enough, for those genreally have drag coefficietns of 1 -- air likes to flow around things. 

calculating a drag coefficient from first principles is dumb.

Anyway, our formula is in the right form, all we need to do is turn a 2 into a 1/4, so an overall adjustment of 1/8 should do it.

14 Pa (high altitude low speed)

1.03e7 Pa (low altitude low speed)

130 Pa (high altitude high speed)

9.29e7 Pa (low altitude high speed)

Comets and Chrondites still break up at high altitude, and Stony and Irons still break up at low altitude. The speed has less bearing than the altitude, likely because of such sharp increases in air density.

# Problem 8.4 \[Back to [top](#toc)\]
$$\label{P8.4}$$

*a) Calculate the size of an iron meteor (density 8$g/cm^3$) that passes through an amount of atmospheric gas equal to its own mass en route to the surface of the Earth. You may assume a spherical meteorite, vertical entry into the atmosphere, and neglect ablation.*

So, first of all, we can set mass equal to mass. Mass of our spherical object:

$$ m = \rho \frac43 \pi r^3 $$

And mass of the atmosphere for a vertical shaft is trivial: its pretty closely approximated by the weight you feel above you! $PA = mg$. The A in this case is the cross sectional area. Thus...

$$ \rho\frac43 \pi r^3 = \frac{P}{g} \pi r^2 $$

And so...

$$ r = \frac{3P}{4g\rho} $$

1 atm is about 1e5 Pa so we have everything we need. (Convert density to 8000 $kg/m^3$). The result is

$$ r = 0.957 m $$

So about a meter. 

*b) Repeat your calculation for a chondritic meteorite of density 4 $g/cm^3$*

This is precisely half the above density, so the radius doubles. r=1.91 m, about 2 meters. 

*c) Repeat your calculation in part a) for an entry angle of 45 degrees.*

So, coming in at 45 degrees, huh? Let's assume a slab atmosphere since even 45 degree angles aren't going to seem like much to the slow curvature of the Earth. So, in a flat atmosphere, what is the scaling factor provided by a 45 degree change? a 90 degree change is suddenly going through infinite atmosphere, so that's no help. However, what we can do is measure the change in *length* of the overall line at every moment. Every "unit" of change that goes straight up is a $\sqrt 2$ "unit" that goes 45 degrees. So the length to the top of the atmosphere is greater by a factor of $\sqrt 2$ and in a plane atmosphere, the same amount of time is spent at each layer, so it's a simple multiplicaiton. This increases the mass on the *atmosphere* side, so our new equation is:

$$ r = \frac{3P\sqrt 2}{4g\rho} $$

Which just means multiply by the factor. 1.35 and 2.71 meters. 

Granted, this is assuming flat atmosphere. This will be bad for large angles of incidence, for instance, but we think it's fine for here. The next order approximation would reduce the size slightly. 

Seemingly magically, the answers are all of the order of meters. Strange how that works out.

The "out" is emission (in IR) and "ref" is reflect (in VIS). We just need to find a way to relate everything together. 

First, consider $P_{sol}$, total intercepted solar power. This would be the solar flux at the asteroid times the cross section. The solar flux at Earth is 1.362 $kW/m^2$, and so using $1/r^2$ scaling to 2.5 AU we find that it has gone down to 0.2179 $kW/m^2$ which converts to 2.179e5 $erg/cm^2s$, which is what we want. Also we have for the reflected portion

$$ P_{ref} = A_b F \pi R^2 $$

So, with all these powers, what do our relations say?
