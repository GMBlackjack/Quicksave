# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

This is quasi-electrostatic--the magnetic field is not changing. (The electric field most certainly is changing due to charge buildup.) 

First we should figure out exactly what the rate of linear charge growth is. It's 2Kt --one Kt from each side reaches the line and stops. (The corrolary is the line current I, stopping at a point and accumulating with It, the only real difference is that we have two currents coming to a single "point" or "line" in this case).

So the line charge itself is $\lambda = 2Kt$, which means $d\lambda/dt = 2K$. Not surprising at all, and there's no need to keep the vector nature of K as its the same everywhere on the line. 

Now, to find the displacement current we first need to find the electric field produced by our "line charge." This is a somewhat well known result. Griffiths 2.9 declares it as

The electric field due to a point charge is

$$ E = \frac{\lambda}{2 \pi \epsilon_0 z}$$

where z is the distance from the wire. Directionality would be radial, but that's not very important at the moment. Substituting in our value for the line charge density...

$$ E = \frac{Kt}{\pi \epsilon_0 z}$$

Changing electric fields produce displacement currents, so we can find it almost directly from here. The derivative of the electric field with respect to time is...

$$ \frac{K}{\pi \epsilon_0 z} $$

Now to get the displacement current we multiply by $\frac{1}{\mu_0 c^2}$

$$ \frac{K}{\pi \mu_0 \epsilon_0 c^2 z} $$

We wish to state explicitly here that $\mu_0 \epsilon_0 = 1/c^2$ just so we don't get confused later.

$$ \frac{K}{\pi z} \hat \rho $$

This is our displacement current! We used $\rho$ instead of r since that's the tradition for cylindrical coordinates. 

The extension of our problem here from previous work suggests that the same Amerian loop we use for the infinite plate we should use here. For a wire, we used a circle. For a plate, we use a rectangular loop. Notably, each loop will have to deal with the displacement current in addition to the current from the plates themselves. Fortunately, we are symmetric along the direction the central line charge lies. 

Ampere's law can be stated

$$ \oint \vec B \cdot d\vec l = \mu_0 I_{enc} $$

In the case of a rectangle, we can state that the sides perpendicular to the plates have no magnetic field in them at all, and that the symmetry in the line charge's direction dictates that the field be constant along the actual paralell lines. (Furthermore, that both the paralell lines have to have the same magnitude of magnetic field, as they are symmetric across the plane.) 

With this understood, let's just consider the part of the amperian loop that's above the plane. 

$$ \int_{y_1}^{y_2} \vec B \cdot d\vec l = \frac12 \mu_0 I_{enc} $$

We consider the direction of the line charge to be y. The planes sit on the xy plane, and z will be a distnace above or below the planes. We do not know the direction of v, but the direction of l we declare to be in the y direction (remember that it will have to be the negative y direction below the plane). Thus, only the y component of B survives this interaction in the dot product. (In fact B has ONLY y components, but we'll argue that later). 

$$ \int_{y_1}^{y_2} B_y dy = \frac12 \mu_0 I_{enc} $$

Unsurprisingly, the integral does not actually depend on the y variable, so we can just replace that with a the length of the loop, L. 

$$ L B_y = \frac12 \mu_0 I_{enc} $$

The current enclosed by the loop is rather simple, LK, with the sign changing based on which side of the line charge we're on. However, the more complicated part of this problem is finding the enclosed current due to the displacement current. 

$$ L B_y = \frac12 \mu_0 (LK + I_{denc}) $$

We implicitly chose a flat surface by declaring the normal I enclosed to be LK, so we must use a flat surface when dealing with the displacement current as well. Since every "slice" along y is identical we know we must multiply the current that pierces along a single "slice" times L to get our results (which is good, as it will cancel out L entirely in the end). Now, we could just use the magnitude of the displacement current to figure this out, but we have what we think is a better idea: find what angle is occupied by the amperean loop, and note that that fraction over 360 degrees will give us the total current through the loop at that slice out of a fraction of 2K. Let the distance away from the line charge be x, and the height z. atan(z/x) will give us the exact fraciton we need here. 

$$ L B_y = \frac12 \mu_0 (LK - \frac{atan(z/x)}{\pi} 2LK) $$

$$ B_y = \frac12 \mu_0 K (1 - 2\frac{atan(z/x)}{\pi}) $$

Note the subtraction, as the displacement current is always going the opposite direction of the sheet current. Also note that when atan is maximized, we end up with 1-1=0, which is what should happen right up close to the line charge. 

Now, we still only have $B_y$ here. Are there any other directional components to the magnetic field? From the plates, definitely not, the field has to be perpendicular to them and it would cancel in the middle. But the displacement current--would it provide anything? 

Well, potentially, but it would have to be in the y direction, as we have perfect symmetry at every y slice. If we pointed anywhere else, that would introduce a divergence into the magnetic field! Thus, we say:

$$ B = \frac12 \mu_0 K (1 - 2\frac{atan(z/x)}{\pi}) $$

Now this is a bit naieve, we now have to be careful with where we are. What we solved for was the case for x>0 and z>0. How do we know this? well, by right-hand-rule and assuming the y direction to be "into" the page on the diagram, the x>0 plate has magnetic field pointing in the positive y direction. Since we had our B be positive, that's where we are. 

Now, the system is symmetric across the xy and yz planes, so we have the solution for all four quadrants of our system. Except, since our relation here depends on both z and x, it is possible a simple "introduction of a minus sign" will ruin everything, so we have to be careful when flipping, since directionality is important. 

Let us call x>0 and z>0 quadrant I.

$$ B_I = \frac12 \mu_0 K (1 + 2\frac{atan(z/x)}{\pi}) $$

Quadrant II will be x<0 z>0. Here the amperian loop's line points in the y direction, but the enclosed current is precisely the opposite of what it was on the other side. A simple "sign change" will only work for the plates, though, as its current is independent of the position of hte line. When x goes negative it flips the sign of the operation, therefore...

$$ B_{II} = \frac12 \mu_0 K (-1 + 2\frac{atan(z/x)}{\pi}) $$

At x=0 there's a discontinuity, but the limit as x approaches 0 maximizes atan to $\pi/2$, which means at x=0 we have a 0 magnetic field. Which is exactly what we should have. 

Now, we consider x>0 z<0, under the plates, as region III. Here, we need to recall that the line element of the amperian loop goes the opposite direction, so that introduces a flip in *both* quantities. Thus in this case a simple negative sign application does the trick.

$$ B_{III} = \frac12 \mu_0 K (-1 - 2\frac{atan(z/x)}{\pi}) $$

And for the last region we just follow the same steps we did for I to II.

$$ B_{IV} = \frac12 \mu_0 K (1 - 2\frac{atan(z/x)}{\pi}) $$

Lastly, we are asked to consider the z=0 case. Since we are literally going through the plate here, of course there would be a discontinuity. As z approaches 0, the displacement current contribution vanishes, leaving us with just +1 and -1 (or -1 and +1). A sharp discontinuity but that's exactly what we expect from plate charges. 

Note: technically we end up with $\pm \frac12 \mu_0 K$ but the leading constants don't really matter to us at this juncture.

With that, we have the magnetic field everywhere and the boundary behavior at the x=0 and z=0 is sensible. We find it quite interesting that the presence of the line charge forces the x=0 boundary to be smooth, while if it were absent there would be a dramtaic shift in the magnetic field, hinting at the fact that trying to ignore accumulating charge is a bad idea and will lead you to the wrong answer.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

The magnetic field of the dipole is something we should be familiar with (but we still had to look up)

$$ \vec B = \frac{\mu_0}{4\pi} \left[ \frac{3(\vec m \cdot \hat r_2)\hat r_2 - \vec m}{r_2^3} \right] $$

And flux is

$$ \phi = \int_S d\vec S \cdot \vec B $$

Though in terms of vector potentials is likely to be easier to do mathematically.

$$ \vec A = \frac{\mu_0}{4\pi} \frac{\vec m \times \vec r}{r^3} $$

And flux is

$$ \Phi = \int_S \vec A \cdot d\vec l $$

We can somewhat easily place the vector potential into the flux integral.

$$ \Phi = \frac{\mu_0}{4\pi} \int_S (\vec m \times \vec r) \cdot d\vec l $$

We know what most of these are. The magnetic moment is in the z direction, the line element is purely angular, but what about the radius? Well, in cylindrical it would be given by $\rho \hat \rho + z' \hat z$ (assuming $z_0$ is zero. We'll add this back in later). 

$$ \Phi = \frac{\mu_0}{4\pi} \int_0^{2\pi} \frac{(m \hat z \times (\rho \hat \rho + z' \hat z))}{r^3} \cdot a d\theta \hat \theta $$

$$  = \frac{\mu_0}{4\pi} \int_0^{2\pi} \frac{(m a \hat \theta)}{r^3} \cdot a d\theta \hat \theta $$

$$  = \frac{\mu_0}{4\pi} \int_0^{2\pi} \frac{(m a)}{r^3} a d\theta $$

$$  = \frac{\mu_0 m a^2}{2 r^3} $$

Which is exactly what we sought.

Of note: we used $\sqrt{\rho^2 + z'^2} = r$ for this, but that's just with the magnet at the origin. Fortunately if we shift for allowing the magnet to be anywhere, $\sqrt{\rho^2 + (z_0 - z')^2} = r$, this changes nothing about the form of our answer!

The EMF is the change in flux over time. Our expression does not feature time, but we can likely extract it. Assuming one-dimensional velocity that doesn't change, in this case v = dz'/dt. So we can re-arrange (in a way that would make mathemeticians scream) to get dt = dz' / v. Keep in mind that in our case v is the *speed* not the *velocity*, and since we are falling downward we need to add a negative sign. 

Which allows for a direct relation between $ -\frac{\partial \Phi}{\partial t} = v \frac{\partial \Phi}{\partial z'}$

Which is what we sought to show. Curiously, we have not specified that the velocity has to be small to find this relation. Wonder why large velocities change this? All it would do is make the flux change faster, wouldn't it? 

After talking about this in class, it is clear that once v gets fast enough we leave the Quasi-Magnetostatic picture entirely and thus simply trying to use Faraday EMF as though we are in a Quasistatic situation is a bad idea. The playing with derivatives above does not change with v, it's the Faraday EMF itself that does. What does it change into? Unimportant, we just know that we shouldn't trust our old equations when we leave quasistatic pictures. 

For the record we can actually evaluate that derivative. 

$$ \frac{\partial \Phi_B}{\partial z'} = \frac{\partial}{\partial z'} \frac{\mu_0 m a^2}{2 r^3} = \frac{\partial}{\partial z'} \frac{\mu_0 m a^2}{2 \sqrt{a^2 + (z_0-z')^2}^3} = \frac{3\mu_0 m a^2(z_0-z')}{2(a^2 + (z_0-z')^2)^{5/2}} $$

Ohm's law is V=IR. The EMF can be thought of as the potential, so we can rewrite this as

$$ I = \frac{3\mu_0 m a^2 v (z_0-z')}{2(a^2 + (z_0-z')^2)^{5/2}} \frac{1}{R} $$
$$ I = \frac{3\mu_0 m a^2 v (z_0-z')}{2r_0^5} \frac{1}{R} $$

Now resistance R is equal to $L/\sigma A$. The length would be the loop of wire, so $L = 2\pi a$. A is the cross sectional area. For an actual look it would be the thickness squared, but we are not in an actual loop, we are in an infinitesimal loop, so the actual "area" is t dz'. 

Which also means our naieve use of "I" is actually "dI" since it's just a small slice of a larger surface current K.

$$ dI = \frac{3\mu_0 m a^2 v (z_0-z')}{2r_0^5} \frac{\sigma t}{2\pi a} dz' $$

$$ dI = \frac{3\mu_0 m a v \sigma t (z_0-z')}{4 \pi r_0^5} dz' $$

Which is what we sought to show. 

First of all, the force dotted with the velocity is just -Fv, as v is speed and not velocity the sign must be moved in front. The sign of F itself is unknown and will be found later. As was discussed in an email, though, the dot product should presumably have absolute value signs around it as we have been dealing with magnitudes this whole time. 

Our integral is not hard to set up at all. 

$$ \int v \frac{\partial \Phi_B}{\partial z'} \frac{3 \mu_0 m a v \sigma t}{4\pi} \frac{z_0 - z'}{r_0^5} dz' $$

We have a lot of constants we can pull out. 

$$ = \frac{3 \mu_0 m a v^2 \sigma t}{4\pi} \int_{-\infty}^{\infty} \frac{\partial \Phi_B}{\partial z'} \frac{z_0 - z'}{r_0^5} dz' $$


While at first it looks like integration by parts would help, that just moves the derivative from the flux to the other half of the equation, and they are quite similar so we might as well just do it manually from this point, as it's not impossible to take the derivatives, just annoying due to all the $r_0$ terms. We have done the derivative in a previous part as well. 

So our full integral becomes...

$$ \frac{3 \mu_0 m a v^2 \sigma t}{4\pi} \int_{-\infty}^{\infty} \frac{3\mu_0 m a^2(z_0-z')}{2(a^2 + (z_0-z')^2)^{5/2}} \frac{z_0 - z'}{(a^2 + (z_0-z')^2)^{5/2}} dz' $$
$$ = \frac{9 v^2 \sigma t \mu_0^2 m^2 a^3}{8\pi} \int_{-\infty}^{\infty} \frac{(z_0-z')^2}{(a^2 + (z_0-z')^2)^{5}} dz' $$

Now this is a decidedly messy integral, but it is also one that is *finite* so it makes sense to what we're currently doing. We will just throw it into an integral solver though. While the indefinite integral is truly nightmarish... (too long to even display, geez) the definite integral is

$$ \frac{5\pi}{128 a^7} $$

And so...

$$ Fv = \frac{9 v^2 \sigma t \mu_0^2 m^2 a^3}{8\pi} \frac{5\pi}{128a^7} $$

$$ = \frac{45 v^2 \sigma t \mu_0^2 m^2}{1024 a^4}$$

And so if we divide out v we get

$$ F = \frac{45 v \sigma t \mu_0^2 m^2}{1024 a^4} $$

Which is the magnitude of the force. Directionally, it should point up. 

Assuming F=w, we get 

$$ v = \frac{1024 a^4 w}{45 \sigma t \mu_0^2 m^2} $$
