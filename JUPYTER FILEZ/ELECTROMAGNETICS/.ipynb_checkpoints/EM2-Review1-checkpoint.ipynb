# The Equations of Electromagnetics

When in doubt, always return to these six equations:

$$ \nabla \cdot \vec E = \frac{\rho}{\epsilon_0} $$

$$ \nabla \times \vec E = -\frac{\partial \vec B}{\partial t} $$

$$ \nabla \cdot \vec B = 0 $$

$$ \nabla \times \vec B = \mu_0 \vec j + \frac{1}{c^2} \frac{\partial \vec E}{\partial t} $$

$$ \vec F = q\vec E + q\vec v \times \vec B $$

$$ \frac{\partial \rho}{\partial t} + \nabla \cdot \vec j = 0 $$

# Quasi-Static Systems

Systems where B is barely changing are Quasi-Electrostatic.

Systems where E is barely changing are Quasi-Magnetostatic. 

The reason it seems backward is that the Quasi-Static systems are named for situations where the static versions of Maxwell's equations hold. 

Some notes:

In Quasi-Electrostatic, the changing E field will cause a displacement current, and the two currents together will be what brings about the magnetic field. 

Another note is that for the quasi-static picture to hold, the change in time is only allowed to be *linear* and sometimes this can be exploited. 

A way to test for "does quasi-static hold?" is to assume it does and then figure out the induced field. If the change in the field can become larger than 1 with specific conditions, you know the conditions that must hold for the picture to be reliable.

Hard to come up with things to review for this one aside from just studying examples, as it's mostly just Magnetostatic and Electrostatic problems to begin with, the trick is identifying which of them to be using. 

# Quasi-Static Situations in Matter (Conductors specifically) 

For conductors $\vec j = \sigma \vec E$. So we can create the following relation.

$$ \frac{\rho}{\epsilon_0/\sigma} = \frac{-\partial \rho}{\partial t} $$

The time constant is $\epsilon_0 / \sigma$. For a Quasi-Electrostatic system this must be large. This is not the case for most conductors.

In Quasi-Magnetostatic, however, the picture does hold, though the relation is not qutie as obvious. 

$$ \nabla \times \vec B = \mu_0 \left[ \frac{1}{\sigma} \vec E + \epsilon_0 \frac{\partial \vec E}{\partial t} \right] $$

One of the primary results of this are Eddy Currents, currents which arise due to changing Em fields. A changing magnetic field produces a current, which is generally found by using B to find E and then using E to find the current j. Eddy currents are stronger at the edges of objects. 

As a reminder, force is...

$$ \vec F = \int \vec j \times \vec B d^3r $$

For AC current, the skin depth exists.

$$ \delta = \sqrt{\frac{2}{\sigma \mu_0 \omega}} $$

# Poynting Vector

The Poynting Vector is given by

$$ \frac{\vec E \times \vec B}{\mu_0} $$

Which represents the direction and magnitude of energy flow. However, the actual rate of energy is somewaht distinct;

$\frac{\partial U}{\partial t} = -\vec j \cdot \vec E - \nabla \cdot \vec S$

It's actually rather simple to calculate most of this, even if it's somewhat hard to think about. 

Individual energies of E and B fields are

$$ \frac{\epsilon_0 \vec E^2}{2}; \frac{\vec B^2}{2\mu_0} $$


# Electromagnetic Angular Momentum

The overall density of electromagnetic angular momentum is

$$ \vec r \times \vec S / c^2 $$

In thory once all EM fields are turned off, the object should start spinning with a torque of

$$ \vec N = \int r \times \rho(\vec r) |vec E(\vec r)d^3r $$

Where $\rho$ is the charge dsitribution. 

# EM Waves in Free Space

In free space the Maxwell equations ahve a simpler form.

$$ \nabla \cdot \vec E = 0 $$

$$ \nabla \times \vec E = -\frac{\partial \vec B}{\partial t} $$

$$ \nabla \cdot \vec B = 0 $$

$$ \nabla \times \vec B =  \frac{1}{c^2} \frac{\partial \vec E}{\partial t} $$

From which come the following differential equations

$$ \nabla^2 \vec E - \frac{1}{c^2} \frac{\partial^2 \vec E}{\partial t^2} = 0 $$

$$ \nabla^2 \vec B - \frac{1}{c^2} \frac{\partial^2 \vec B}{\partial t^2} = 0 $$

Which have solutions

$$ \vec E = \vec E_0 cos(\vec k \cdot \vec r - \omega t + \phi_0) $$
$$ \vec B =  \frac{\vec k}{\omega} \times \vec E_0 cos(\vec k \cdot \vec r - \omega t + \phi_0) $$

In generality, the energy density is

$$ \langle u_{EM} \rangle = \frac{\epsilon_0 E_0^2}{2} $$

$$ \langle \frac{1}{\mu_0 c} E_0^2 \hat k \rangle $$

We can also write everything in complex notation.

$$ \vec E = \vec E_0 e^{i(\vec k \cdot \vec r - \omega t)} $$
$$ \vec B =  \frac{\vec k}{\omega} \times \vec E_0 e^{i(\vec k \cdot \vec r - \omega t)} $$

Though only the real parts of these are actually what we want. 

Polarization is determined by the direction of $\vec E_0$. If it is real, it is planar. If it is complex, it is some kind of eliptical polarizaiton. Special cases are circular, though left and right circular aren't really specified. ($\hat x - i \hat y, \hat x + i \hat y$). 

Phase speed: $\omega/k$ for this. Which is c. 

Most of the math involving these waves is simply additive, though extracting exponentials will be a skill that serves you well. 

# EM Waves in Linear Media

In matter we represent Maxwell's equations as

$$ \nabla \cdot \vec D = \rho_f $$

$$ \nabla \times \vec E = -\frac{\partial \vec B}{\partial t} $$

$$ \nabla \cdot \vec B = 0 $$

$$ \nabla \times \vec H = \vec j_f + \frac{\partial \vec D}{\partial t} $$

Though we often have cases where there is no free charge and no current.

$$ \nabla \cdot \vec D = 0 $$

$$ \nabla \times \vec E = -\frac{\partial \vec B}{\partial t} $$

$$ \nabla \cdot \vec B = 0 $$

$$ \nabla \times \vec H = \frac{\partial \vec D}{\partial t} $$

For linear media, $\vec D = \epsilon \vec E; \vec B = \mu \vec H$ which allows us to say 

$$ \nabla^2 \vec E - \mu\epsilon \frac{\partial^2 \vec E}{\partial t^2} = 0 $$

$$ \nabla^2 \vec H - \mu\epsilon \frac{\partial^2 \vec H}{\partial t^2} = 0 $$

Similarly for the D and B fields. In matter, $\vec S = \vec E \times \vec H$. 

And the waves can be represented in exponential form

$$ \vec E = \vec E_0 e^{i(\vec k \cdot \vec r - \omega t)} $$
$$ \vec H = \vec H_0 e^{i(\vec k \cdot \vec r - \omega t)} $$

But in this case we have some inter3esting relations. For instance, $k = n\omega / c$ and $\hat k \times \vec E = \sqrt{\frac{\mu}{\epsilon}} \vec H$, and $Z = \sqrt{\frac{\mu}{\epsilon}}$. We also note the index of refraction n is $n = \sqrt{\frac{\mu \epsilon}{\mu_0\epsilon_0}}$. Most materials have the $\mu$ values comparable.

The funny thing is that n can be different based on polarizaiton, adjusting different "components" of the EM wave separately. 

# Basic Optics

When going from one medium to another, there are boundary conditions.

$$ \hat n \cdot (\vec D_1 - \vec D_2) = 0 $$
$$ \hat n \times (\vec E_1 - \vec E_2) = 0 $$
$$ \hat n \cdot (\vec B_1 - \vec B_2) = 0 $$
$$ \hat n \times (\vec H_1 - \vec H_2) = 0 $$

From these we can develop a lot of things. There's a lot of proofs involved but the main results are the familiar ones...

$$sin\theta_I = sin\theta_R$$
$$ n_1sin\theta_I = n_2sin\theta_T $$

But the more itneresting results are the relations of the magnitudes of the transmitted and reflected wabes, because they depend on whether we are Transverse Electric (E comes out of the board) or Transverse Mangetic (B comes out of the board). 

TM says

$$ \frac{E_R}{E_I} = \frac{Z_1 cos\theta_1 - Z_2 cos\theta_2}{Z_1 cos\theta_1 + Z_2 cos\theta_2} $$
$$ \frac{E_T}{E_I} = \frac{2Z_2 cos\theta_1}{Z_1 cos\theta_1 + Z_2 cos\theta_2} $$

TE says

$$ \frac{E_R}{E_I} = \frac{Z_2 cos\theta_1 - Z_1 cos\theta_2}{Z_1 cos\theta_1 + Z_2 cos\theta_2} $$
$$ \frac{E_T}{E_I} = \frac{2Z_2 cos\theta_1}{Z_1 cos\theta_1 + Z_2 cos\theta_2} $$ 

One may argue about energy in must equal energy out. Well, this is true for the poynting vector. What we need to check is the *power*, specifically, that is $\frac{E_R^2}{E_I^2} \frac{Z_1 cos\theta_1}{Z_1 cos \theta_1} + \frac{E_T^2}{E_I^2} \frac{Z_1 cos\theta_2}{Z_2 cos \theta_1}$.
