# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

So we can just assume quasi-magnetostatic is true, this makes things somewhat easier. 

In quasi-magnetostatics, $\frac{\partial E}{\partial t} << 1$ so we might as well consider it 0 in most cases (but we should be careful). 

First, let's find the electric field. Let's just go to Maxwell's Equations.

$$ \nabla \times \vec B = -\mu_0 \vec j $$

This in particular is not helpful. The magnetic field does not depend on any physical coordinates at all, so its cross product is zero everywhere. So we turn to a different relation.

$$ \nabla \times \vec E = -\frac{\partial \vec B}{\partial t} $$

Take the derivative...

$$ \Rightarrow \nabla \times \vec E = -B_0 \omega cos(\omega t) \hat z $$

Only the z direction remains, so this gives us some information about the cross product. 

$$ \Rightarrow \frac{\partial E_y}{\partial x} - \frac{\partial E_x}{\partial y} \hat z = -B_0 \omega cos(\omega t) \hat z $$

Which, while somewhat expected, isn't the most helpful. Instead we use the cylindrical coordinates definition. We have azimuthal symmetry and we can't have any charge buildup, so there can only be angular electric fields, no radial. Thus...

$$ \Rightarrow \frac{1}{\rho} \frac{\partial (\rho E_\phi) }{\partial \rho} \hat z= -B_0 \omega cos(\omega t) \hat z $$

While there can't be any radial componetns in E, there can certainly be a radial dependence, so the derivative is not trivial. Try to simplify...

$$ \Rightarrow  \frac{1}{\rho}E_\phi + \frac{\partial E_\phi }{\partial \rho} = -B_0 \omega cos(\omega t) \hat z $$

Which is a solvable differential equation if E depends linearly on $\rho$. 

The solution to this differential equation is 

$$ E_\phi = -\frac{1}{2}B_0\omega cos(\omega t)\rho + \frac{C}{\rho} $$

Now the integration constant can be set to zero since there can't be a term exploding at the center of the cylinder. 

$$ \vec E = -\frac{1}{2}B_0\omega cos(\omega t)\rho \hat \phi $$

Larger electric field at the edges of the cylinder means larger current.

$$ \vec j_0 = - \sigma \frac{1}{2}B_0\omega cos(\omega t)\rho \hat \phi $$

Which are the eddy currents. Stronger at the edges, as they should be. 

The Poynting Vector is given by

$$ \vec S = \frac{\vec E \times \vec B}{\mu_0} $$

Which we can actually calculate directly at the edge of the cylinder.

$$ \vec S = - \frac{1}{\mu_0} \frac12 B_0^2 \omega sin(\omega t) cos(\omega t) R (\hat \phi \times \hat z)$$
$$ -\frac{1}{\mu_0} \frac12 B_0^2 \omega sin(\omega t) cos(\omega t) R \hat \rho $$

By default energy is flowing inward--but we note the presence of the trig functions which will flip the sign regularly. This indicates that the time average *of* the Poynting vector is actually zero, since the integral over a single period of a sincos function is zero. 

We are asked for the flux though, which is just the vector times the surface area of the cyildiner's edge. 

$$ -\frac{1}{\mu_0} B_0^2 \omega sin(\omega t) cos(\omega t) R^2 \pi L\hat \rho  $$

So, moving on to our energy flow. We have: 

$$ \frac{\partial U}{\partial t} = -\vec j \cdot \vec E - \nabla \cdot \vec S $$

So there is certainly energy flow at any given time. However, again, we look for the time average. The physical derifatives of the Poynting Vector do not alter its time-average behavior, so it goes to zero. However, the same is not true of the current and electric field component of this.

$$ \Rightarrow \frac{\partial U}{\partial t} = (\sigma \frac{1}{2}B_0\omega cos(\omega t)R \hat \phi) \cdot (-\frac{1}{2}B_0\omega cos(\omega t)R \hat \phi) $$
$$ \Rightarrow \frac{\partial U}{\partial t} = - \sigma \frac14 B_0^2 \omega^2 cos^2(\omega t) R^2 $$

Now the time average of cosine squared is not zero, it is 1/2. 

$$ \Rightarrow \langle \frac{\partial U}{\partial t} \rangle = - \sigma \frac18 B_0^2 \omega^2 R^2 $$

Where the negative sign clearly indicates flow inward with respect to the surface normal. So even with an oscillating Poynting Vector, the energy still flows inward. 

We had assumed quasi-magnetostatic, that is, $\frac{\partial E}{\partial t} = 0$. This obviously isn't true, for

$$ \frac{\partial E}{\partial t} = \frac{1}{2}B_0 \omega^2 sin(\omega t) \rho \hat \phi. $$

So clearly we get a correction here. 

Now, curiously, we noted in the previous part that $\nabla \times \vec B = - \mu_0 \vec j$ was useless since the magnetic field here has no cross product. Well, now it IS useful, because that was if we had suppressed the electric field term, the full relation is $\nabla \times \vec B = - \mu_0 \vec j + \frac{1}{c^2} \frac{\partial \vec E}{\partial t}$, which we can use to realte:

$$ \vec j_1 = \frac{1}{\mu_0 c^2} \frac{1}{2}B_0 \omega^2 sin(\omega t) \rho \hat \phi $$

This is the current correction. Let's see how it compares with the original...

$$ \vec j_0 = - \sigma \frac{1}{2}B_0\omega cos(\omega t)\rho \hat \phi $$

Similar shape, but the correction depends on the square of the frequency and doens't rely on the conductivity at all. Also note the difference in sign--the new current density fights back! (but is still stronger at the edges). 

This in turn creates a correction in the electric field, though. (Not the magnetic, surprisingly, as the cross product still remains zero overall). 

$$ \vec E_1 = \frac{1}{\sigma \mu_0 c^2} \frac{1}{2}B_0 \omega^2 sin(\omega t) \rho \hat \phi $$

For the Poynting Flux we do need to find the next correction for the magnetic field. For that, we shall use:

$$ \nabla \times \vec E = -\frac{\partial \vec B}{\partial t} $$

The electric field depends only on rho and exists only in the phi direction, so this reduces to

$$ \frac{1}{\rho} \frac{\partial}{\partial \rho}\left( \frac{1}{\sigma \mu_0 c^2} \frac{1}{2}B_0 \omega^2 sin(\omega t) \rho^2\right) \hat z= -\frac{\partial \vec B}{\partial t} $$
$$ \Rightarrow  \frac{1}{\sigma \mu_0 c^2} B_0 \omega^2 sin(\omega t) \hat z = -\frac{\partial \vec B}{\partial t} $$
$$ \Rightarrow  -\frac{1}{\sigma \mu_0 c^2} B_0 \omega cos(\omega t) \hat z = \vec B_1 $$

This is still not dependent on any physical variable, but it does vary out of phase with the other magnetic field. So the correction will give us a different Poynting vector:

$$ \vec S = \frac{\vec E \times \vec B}{\mu_0} $$

However, these are not just the correction terms we deal with, these are the total electric and magnetic fields. Which... are going to be a little complex to keep track of.

$$ \vec E = -\frac{1}{2}B_0\omega cos(\omega t)\rho \hat \phi + \frac{1}{\sigma \mu_0 c^2} \frac{1}{2}B_0 \omega^2 sin(\omega t) \rho \hat \phi $$

$$ \vec B = B_0 sin(\omega t) \hat z - \frac{1}{\sigma \mu_0 c^2} B_0 \omega cos(\omega t) \hat z $$

$$ \Rightarrow \vec S = \left( -\frac12 B_0^2 \omega cos(\omega t) sin(\omega t) R +\frac{1}{2\sigma \mu_0 c^2} B_0^2 \omega^2 cos^2(\omega t) R + \frac{1}{2 \sigma \mu_0 c^2} B_0^2 \omega^2 sin^2(\omega t)R - \frac{1}{2\sigma^2 \mu_0^2 c^4} B_0^2 \omega^3 cos(\omega t) sin(\omega t) R \right) \hat \rho $$

Which we can just multiply by $2 \pi R L$ to get the total flux. 

$$ \left( -\frac12 B_0^2 \omega cos(\omega t) sin(\omega t) R +\frac{1}{2\sigma \mu_0 c^2} B_0^2 \omega^2 cos^2(\omega t) R + \frac{1}{2 \sigma \mu_0 c^2} B_0^2 \omega^2 sin^2(\omega t)R - \frac{1}{2\sigma^2 \mu_0^2 c^4} B_0^2 \omega^3 cos(\omega t) sin(\omega t) R \right) 2 \pi R L \hat \rho $$

Now the *time average* of the Poynting Vector is interesting! The square trig functions evaluate to 1/2, everything else to 0.

$$ \left( +\frac{1}{4\sigma \mu_0 c^2} B_0^2 \omega^2 R + \frac{1}{4 \sigma \mu_0 c^2} B_0^2 \omega^2 R \right) \hat \rho $$

$$ = \frac{1}{2 \sigma \mu_0 c^2} B_0^2 \omega^2 R \hat \rho $$

So now we consider the energy flow.

$$ \frac{\partial U}{\partial t} = -\vec j \cdot \vec E - \nabla \cdot \vec S $$

Looks like we substituted $\rho=R$ too early. No matter, we can still take the divergence with respect to R, it changes nothing. The current contribution though:

$$ -\vec j \cdot \vec E  $$
$$ = -(\frac{1}{\mu_0 c^2} \frac{1}{2}B_0 \omega^2 sin(\omega t) \rho \hat \phi- \sigma \frac{1}{2}B_0\omega cos(\omega t)\rho \hat \phi)(-\frac{1}{2}B_0\omega cos(\omega t)\rho \hat \phi + \frac{1}{\sigma \mu_0 c^2} \frac{1}{2}B_0 \omega^2 sin(\omega t) \rho \hat \phi) $$ 


Only the square trig functions will have a result over the time average. 
 
$$ = - \frac{1}{8\sigma\mu_0^2c^4} B_0^2 \omega^4 R^2 - \frac{\sigma}{8} B_0^2 \omega^2 R^2 $$

And so...

$$ \langle \frac{\partial U}{\partial t} \rangle = - \frac{1}{8\sigma\mu_0^2c^4} B_0^2 \omega^4 R^2 - \frac{\sigma}{8} B_0^2 \omega^2 R^2 - \frac{1}{2 \sigma \mu_0 c^2} B_0^2 \omega^2 $$

This answer is many things. Ugly. A mess. A bit esoteric as to what it actually means. BUt one thing we can say is that there most certainly is an energy flow due to the Poynting Vector, it does not cancel out this time. Also every term is uniformly negative, so energy is continually being lost. Which is sensible as the magnetic field is continually being supplied energy to maintain its driving force.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

First of all, be careful of what's *not* labeled here. An infinite solenoid with a constant magnetic field is driven by a surface current. Since the magnetic field goes in the z direction, this surface current needs to go in the $\phi$ direction. $ B = \mu_0 K$. in this case, $\vec K = B/\mu_0 \hat \phi$. 

This may not be necessary to solve the problem, but these situations are notorious for hiding things we need to consider.

Since the overall charge density has to be equal inside and outside the cylinder, we can relate $\rho$ to $\sigma$. In a slice, we have charges of $2\pi a \sigma = - \pi a^2 \rho$, or $\rho = - \frac{2}{a} \sigma$

Now that we have everything we might want, what is the electromagnetic angular momentum? Well, its density is...

$$ \vec r \times \frac{\vec S}{c^2} $$

Integrating this over all space will get the total angular momentum from electromangetic contributions. 

$$ \vec S = \frac{\vec E \times \vec B}{\mu_0} $$

We have B, we just need to find the electric field. 

The electric field is not that difficult to calculate. We can use a cylindrical Gaussian surface for this. The entire system is electrically neutral, so we have no electric field outside the solenoid. (This does lead us to the somewhat curious question of "what exactly is driving the current?" but let's leave that aside in case it becomes an issue. Since all of it is neutral, most of the inside of the solenoid has no electric field either. However, once we *get* to the cylinder, the electric field points outward based on the total enclosed charge. 

For line charges, the genreal electric field is

$$ \frac{\lambda}{2\pi \epsilon_0 r} \hat r $$ 

where $\lambda$ is the line charge density. In our case, we effectively don't care about the surface charge since its infinitesimal (and must be responsible for the sudden drop to E=0 outside the cylinder). All we care about is the volume charge enclosed on a slice, which is $\pi r^2 \rho$ depending on what r we are at. 

Replacing $\lambda$ with this gives us...

$$ \vec E = \frac{r \rho}{2 \epsilon_0} \hat r $$

Which is sensible, as it gets stronger the further away we get. Note that we are using the r coordinate rather than $\rho$ since we have a $\rho$ density. r is still the cylindrical radius. Note that the E-field only actually exists inside the cylinder, so we only have a Poynting vector inside the cylinder. 

$$ \vec S = \frac{1}{\mu_0} \frac{r \rho B}{2 \epsilon_0} (\hat r \times \hat z)$$
$$ = - \frac{1}{\mu_0} \frac{r \rho B}{2 \epsilon_0} \hat \phi$$

Now we just cross this with the generic raidal vector which, since we only care about a slice, is just $r\hat r$. This means the electromagnetic angular momentum is...

$$ - \frac{1}{\mu_0\epsilon_0 c^2} \frac{r^2 \rho B}{2} \hat z$$
$$ = - \frac{r^2 \rho B}{2} \hat z$$

It really is nice how the speed of light just cancels out here.

Anyway, the electromagnetic angular momentum is oriented "spin down." At least it poitns in the z direction, which was pretty necessary for the geometry of the problem. To get the total, just integrate the radius from 0 to a. 

$$ \int -\frac{r^2 \rho B}{2} \hat z r dr d\phi $$
$$ = \pi \rho B \int - r^3 \hat z dr $$
$$ = - \pi \rho B \frac{a^4}{4} \hat z $$

$$ dL_{EM} = - \pi \rho B \frac{a^4}{4} \hat z $$

Keep in mind this is per unit length. Multiply by L if wanted for an entire section. 

Changing the magnetic field definitely results in a change in the electric field. Specifically, the field it produces is::

$$ \nabla \times E_i = -\dot B$$

Where the time derivative of B is given in dot notation since we are treating it as a variable we express this in terms of.

The magnetic field poitns in the z direction. If we were increasing the field we would be going clockwise, so this is technically in the $\phi$ direction, though the magnetic field derivative itself will be negative and flip this later.

$$  \frac{1}{r} \frac{\partial}{\partial r} (r E_{i \phi}) \hat z = -\dot B \hat z $$

This is very similar to the differential equation we solved in **Problem 1**. Following that same pattern, we get...

$$ \vec E_i = -\frac{1}{2}\dot B r \hat \phi $$

Now, if our cylinder wasn't electrically neutral, this would definitely produce a torque. But it IS electrically neutral... so one side must be pulled on more than another. 

The question is, how do we find torque of E on the charge? Equation 3.4 gives us...

$$ \vec N = \int  \vec r \times \rho(\vec r) \vec E(\vec r) d^3r $$

Note: the B-field has no effect on the torque since originally the cylinder won't be moving, and once it is moving it's only spinning, and thus moving in all directions so any possible contriptions from the B-field will cancel out uniformly across the circle. 

We will naturally need to split this up into two integrals and then add them due to the difference in charge distribution. Also, we should write $\rho$ in terms of $\sigma$ so we can correlate them later.

$$ \vec N_1 = \int  \vec r \times \frac{2\sigma}{a} \frac{1}{2}\dot B r \hat \phi d^3r $$
$$ \vec N_2 = -\int  a \hat r \times \sigma \frac{1}{2}\dot B a \hat \phi d^3r $$

Doing the cross product...

$$ \vec N_1 = \int  r \frac{2\sigma}{a} \frac{1}{2}\dot B r \hat z d^3r $$
$$ \vec N_2 = -\int  a \sigma \frac{1}{2}\dot B a \hat z d^3r $$

Because we are dealing with a cross sectional area torque, we don't take the integral over the volume but just a slice.

$$ d\vec N_1 = \int  r \frac{2\sigma}{a} \frac{1}{2}\dot B r \hat z d^2r $$
$$ d\vec N_2 = -\int  a \sigma \frac{1}{2}\dot B a \hat z d^2r $$

Let's do the central density first.

$$ d\vec N_1 = \int \frac{2\sigma}{a} \frac{1}{2}\dot B r^2 \hat z d^2r $$

When integrating over a circle, the integration factor $d^2r = rdrd\phi$.

$$ d\vec N_1 = \int \frac{2\sigma}{a} \frac{1}{2}\dot B r^3 \hat z drd\phi $$

We have no angular dependence so the angle comes out as $2\pi$ 

$$ d\vec N_1 = 2\pi \frac{\sigma}{a} \dot B \int  r^3 \hat z dr$$

And the integral from 0 to a is...

$$ d\vec N_1 = 2\pi \frac{\sigma}{a} \dot B \frac{a^4}{4} \hat z $$

Which simplifies to 

$$ d\vec N_1 = \pi \sigma \dot B \frac{a^3}{2} \hat z $$

And now for the surface charge.

$$ d\vec N_2 = -\int  a \sigma \frac{1}{2}\dot B a \hat z r dr d\theta $$

Becasue the charge only exists on the edge, r is in a delta function and can be pulled out as a. 

$$ d\vec N_2 = -a^3 \sigma \frac{1}{2}\dot B \int \hat z d\theta $$

The angle integrates freely to $2\pi$

$$ d\vec N_2 = -a^3 \sigma \dot B \pi \hat z$$

And so the total torque per cross sectional area...

$$ d\vec N = \pi \sigma \dot B \frac{a^3}{2} \hat z -a^3 \sigma \dot B \pi\hat z$$
$$ d\vec N = - \pi \sigma \dot B \frac{a^3}{2} \hat z $$

Which is our answer. 

The torque is equal to the derivative of mechanical angular momentum. In our notation, $d\vec N = \frac{d}{dt} dL_{mech}$

So if we integrate with respect to time until the magnetic field goes to 0...

$$ dL_{mech} = -\pi \sigma \frac{a^3}{2} \int \dot B \hat z dt  = \pi \sigma \frac{a^3}{2} B \hat z $$

Since $\int \dot B = - B$, as we are adding up an overall nevative change over time. 

For part a we had

$$ dL_{EM} = - \pi \rho B \frac{a^4}{4} \hat z $$

Remembering that $\rho = -\frac{2\sigma}{a} $ we get...

$$ dL_{EM} = \pi \frac{a^3 \sigma B}{2} \hat z $$

And the two answers are identical, so we are done!
