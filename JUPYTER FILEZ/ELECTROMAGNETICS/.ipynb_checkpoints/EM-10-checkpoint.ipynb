# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Now obviously equilibrium is pointing the same direction as the other dipoles, since that's what dipoles *do*, but we should calculate it as instructed. It will not depend on D. 

We will note that there is no specificity of if these are magnetic or electric moments. That said we will suppose that this is magnetic seeing as we're in the magnetic section of the class. In this case, we can use:

$$\vec N = \vec m \times \vec B$$

If we kept the dipole generic we could prove the final direction from vectors alone, but we'll need a number for part b) so let's construct the actual magnetic field. Which we need to know everywhere around the dipole moment since we are somewhat concerned with how the field *changes* at that point. It will be rather annoying to work with the B-field so instead we use the A-field, at least for now. Each individual A-field is given by:

$$ \vec A = \frac{\mu_0}{4\pi r_2^3} \vec m \times \vec r_2 $$

Fortunately for our point of interest, the magnitude of $r_2$ is the same for both of our "source" dipoles. It's the cross product that changes. A cross product which, when evaluated in components, reads: 

$$ mr_z \hat x - mr_x \hat z $$

The $r_x$ portion remains the same regardless of which dipole we are considering. However, the $r_z$ portion will flip due to the different positions, meaning only the $r_x$ term remains.

Unfortuantely we can't simplify yet since we have to actually use the B-field and technically speaking that involves taking a curl. At least we know the field makes sense. Regardless, we also know B in dipole form, it's just a bit more annoying to look at.

$$ \vec B = \frac{\mu_0}{4\pi} \left[ \frac{3(\vec m \cdot \hat r_2)\hat r_2 - \vec m}{r_2^3} \right]$$

Probably helpful to actually find the radial unit vector.

$$ \hat r = (\frac{D}{\sqrt{d^2+D^2}},\pm\frac{d}{\sqrt{d^2+D^2}},0) $$

We can break up the individual components. 

$$ \Rightarrow \vec B = \frac{\mu_0}{4\pi} \left[ \frac{ \pm 3m\frac{d}{\sqrt{d^2+D^2}} \hat r_2 - m\hat y}{r_2^3} \right]$$
$$ \Rightarrow \vec B = \frac{\mu_0}{4\pi} \left[ \frac{ \pm 3m\frac{dD}{d^2+D^2} \hat x + 3m\frac{d^2}{d^2+D^2} \hat y - m\hat y}{r_2^3} \right]$$

Now if we add both fields together it's clear that the plus and minus term cancels out. So our *whole* field here is:

$$ \Rightarrow \vec B = \frac{\mu_0}{2\pi} \left[ \frac{ 3m\frac{d^2}{d^2+D^2} - m}{(D^2+d^2)^{3/2}} \right]\hat y$$

Now we need to be careful, those moment m terms are the source dipoles. While the magnitude is the same, the direction is different, so when we do $\vec m \times \vec B$ we need to use an uncertain direction (it won't point in z, though, that much we can say.) Since our magnetic field is only in the y direciton we shall write it as $B_y$ to make it clear what's going on.

$$ \vec m \times \vec B = 0 \hat x +  0 \hat y + (m_x B_y - m_z B_y )\hat z = m_x B_y \hat z $$

So we rotate counterclockwise if the x-portion of the moment is positive, and clockwise if it is negative. Thus, as expected, the equilibrium position is when the dipole is pointing striaght up, in the y direction, for no torque is experienced by such a moment. 

Curiously, while D does not change which positions are equilibrium (+y or -y as usual) we can alter D to make the overall sign of $B_y$ change. This just means the dipole will rotate opposite directions depending on where it is. (Absolutely fascinating, to be sure, but it also makes sense. Near the dipoles the field is predominantly upward, while far away it is predominantly downward).

In fact, this implies that there exists a point where the fields exactly cancel to nothing. If we can set $d^2/(d^2+D^2)$ to 1/3, we can do it. Thus, $D^2 = 2d^2 \rightarrow D = d \sqrt{2}$

At this point there is no equilibrium. So we suppose this technically means that there *is* a point where equilibrium depends on D, as there is a point with none.

Note: we are considering -y and +y to both be eqiulibrium situations no matter what D is, as we do not know the initial orientation of the dipole either one is valid depending on where the pointing starts. Which *one* of these is the one a specific dipole seeks does depend on if D is greater than or less than the "no eqilibrium" point. 

We have a function for the torque. 

$$ N = m_x B_y $$

Which we can correlate to the angle.

$$ N = m sin\theta B_y $$

We have decided to use sine and just define the zero angle as straight up. This is so we can use the small angle approximation. Torque also happens to equal $I\alpha$

$$ \alpha = m sin\theta B_y / I $$

With the small angle approximation we can write this as a differential equaiton.

$$ \theta'' = m \theta B_y / I $$

This differential equation is a known form with a known solution, something of the form Asin(at)+Bcos(at). At the ends of oscillation the moment better not still be moving, so we remove the sine term since its derivative would be positive at t=0, and we can freely set the boundary conditions to actually *be* zero. Therefore...

$$ \theta = A cos(\sqrt{\frac{mB_y}{I}}t) $$

Now the amplitude A here depends on where exactly we start the oscillations. Fortunately we do not care, we can just read off the period from the constant inside the cosine term.

$$ T = 2\pi\sqrt{\frac{mB_y}{I}}  $$

Which is our answer. 

Yes, this was based off the proof for the pendulum period. Notably we did try to do it without resorting to a differential equation first, but that ended up in the weeds quickly trying to extract time dependence from an angle so we could take the integral with respect to time OF the angle...

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

Note: problem has an issue, force should be negative. Our goal is to *have* the minus sign. 

So, "use an energy method" would imply refering to the Energy in Magnetostatics section. Which means we have to use the relation::

$$ \delta \hat V_{21} = - \vec F_{21} \cdot \vec d $$

We do have to be quite careful with the sign here, as there's the "hidden electric field nonsense" that raises its ugly head. V is the interaction energy. We can, in fact, solve for that, for $ V_{21} = -I_1 \Phi_{12} $ The flux through the wire doesn't make much sense, but the flux through the *loop*, that we can figure out. 

The flux through the loop (which we shall notate as object "1") due to the wire ("2") can be calculated from the genreal integral for magnetic flux:

$$ \Phi_{12} = \oint_S \vec A_2 \cdot d\vec l $$

The vector potential for a wire is well known. 

$$ \vec A_2 = \frac{\mu_0 I_2}{2\pi} ln(L/2r) \hat x $$

Notably this implies that the wire has a length L. But this L is considered "really big." (the only reason the L needs to be here is that in the derivation, magnetic fields diverge for an infinite wire. Joyous. It vanishes when the cross product is taken to find B, but we might not be doing that here, we'll have to wait and see). 

Regardless, we have to be careful, that "r" is distance from the wire, not distance from the center of the circle. If we want to know A at the points on the loop, we need to redefine it in terms of angle. Which we can actually do! $r = d + Rsin\theta$. With this, we write...

$$ \Phi_{12} = \oint_S \frac{\mu_0 I_2}{2\pi} ln(\frac{L}{2(d+Rsin\theta)}) \hat x \cdot R d \hat \theta $$

Now we currently have mixed coordinates, but we need to dot the angle direction with the x direction. The result is dependent on angle, specifically $-sin\theta$. The minus sign is there because the angle that points in the same direction as x is at the bottom of the circle, or $\theta = 3\pi/2$. With this, we now have...

$$ \Phi_{12} = - \oint_0^{2\pi} \frac{\mu_0 I_2}{2\pi} ln(\frac{L}{2(d+Rsin\theta)}) R sin\theta d \theta $$
$$ \Rightarrow \Phi_{12} = -  \frac{\mu_0 I_2 R}{2\pi} \oint_0^{2\pi} ln(\frac{L}{2(d+Rsin\theta)}) sin\theta d \theta $$

Now this. This is an ugly integral. We can simplify it somewhat by splitting up the natural log. 

$$ \Rightarrow \Phi_{12} = -  \frac{\mu_0 I_2 R}{2\pi} \oint_0^{2\pi} \left( ln(L/2) - ln(d+Rsin\theta)) \right) sin\theta d \theta $$

The full integral over sine (or cosine for that matter) for all angles goes to zero. So we can get rid of L--as expected, it does not matter in the slightest.
 
$$ \Rightarrow \Phi_{12} = \frac{\mu_0 I_2 R}{2\pi} \oint_0^{2\pi} ln(d+Rsin\theta)) sin\theta d \theta $$

This we have the computer examine.



Naturally x is $\theta$ in this case, and wolfram rewrote ln and log. Truly ugly. However, we are evaluating a *definite* integral, which simplifies things considerably. However, we notice that there is a *pole* hiding in the result here, a discontinuity that would ruin everything. Fortunately we can redefine our bounds for the cirlces, from $[0,2\pi]$ to $[-\pi, \pi]$, which will produce something. 

We do need to be careful though, the tangetnt of $\pm \pi/2$ diverges. However, it diverges INSIDE of an arctan function, which produces a finite number in the limit. The cosine term evaluates to zero, while the x term does not. So we end up with...

$$ \Rightarrow \Phi_{12} = \frac{\mu_0 I_2}{2\pi} \left[ -2 \sqrt{d^2-R^2}\pi + 2d\pi \right] $$

Then we evaluate and simplify...

$$ \Rightarrow \Phi_{12} = \mu_0 I_2 \left[ - \sqrt{d^2-R^2} + d \right] $$

Which would make

$$  V_{21} = -\mu_0 I_1 I_2 \left[ - \sqrt{d^2-R^2} + d \right] $$

This is the interaction energy. Of course we actually need the *force* which is going to require taking the derivative with respect to the distance d...

$$ " F" = -\mu_0 I_1 I_2 \left[ - \frac{d}{\sqrt{d^2-R^2}} + 1 \right] $$

And remember, we have to flip the sign! (Hence the quotes around the "force" above).

$$ F = -\mu_0 I_1 I_2 \left[ \frac{d}{\sqrt{d^2-R^2}} - 1 \right] $$

Which is exactly what we wanted to find. Though, admittedly, this is just the magnitude. Fortunately the direction is very easy to tell from context: the magnetic field is azimuthally symmetric and so the only gradient possible is along the radius, which would be in the z direction here. 

$$ \vec F = -\mu_0 I_1 I_2 \left[ \frac{d}{\sqrt{d^2-R^2}} - 1 \right] \hat z$$

$$ \vec F = -\mu_0 I_1 I_2 \left[ \frac{d}{\sqrt{d^2-R^2}} - 1 \right] \hat z $$

Now in the first order approximation this just goes to zero since d/d-1=0. So we need to expand and for expansions we need to work with 1 >> R/d. So we rewrite our relation...

$$ = -\mu_0 I_1 I_2 \left[ \frac{d}{d\sqrt{1-R^2/d^2}} - 1 \right] \hat z $$
$$ = -\mu_0 I_1 I_2 \left[ \frac{1}{\sqrt{1-R^2/d^2}} - 1 \right] \hat z $$

Now, rather than expanding the entire funciton (which will be somewhat awkward) we are going to expand the fractional term by itself.

$$ \frac{1}{\sqrt{1-R^2/d^2}} \rightarrow 1 + \frac12 \frac{R^2}{d^2} + O(R^4/d^4) $$

Anyway if we substitute the two terms back into the original function, we see that of course the first order vanishes in the subtraction, leaving the squared correciton term

$$ \vec F = -\mu_0 I_1 I_2 \left[ \frac12 \frac{R^2}{d^2} \right] \hat z $$

This is for large d when compared to R.

Now, we seek to compare this to the known result.

$$ \vec F = \nabla (\vec m \cdot \vec B) $$

Naturally our moment is the loop, which is pointing in the z direction (out of the page) so the magnetic field better also be pointing the same direction. Which it is, as that's where the loop is in relation to the wire. So the dot product becomes a regular product, and since the magnetic moment is constant, the relation is now

$$ \vec F = m \nabla B \hat z  $$

Magnetic field for a wire is relatively well known.

$$ \vec F = m \nabla \frac{\mu_0I_2}{2\pi r} \hat z \approx - m \frac{\mu_0I_2}{2\pi d^2} \hat z $$

Note the approximately, for the "motoin" here is along the z axis, NOT along d, which is on the y axis. However, for a slight change in the force/energy, the distance from the wire might as well just be d, and so it is, espeically when d>>R. We aren't looking for rigid calculations here anyway, we're jsut trying to compare the magnetic moment definition to the definition obtianed in the previous part. Both of them go like 1/d^2 which is a very good sign. 

But let's go ahead and add the magnetic moment of the loop, see what we get.

$$ \vec F \approx - I_1 \pi R^2 \frac{\mu_0I_2}{2\pi d^2} \hat z = - I_1 R^2 \frac{\mu_0I_2}{2 d^2} \hat z  $$

Fascinating! We were just lookign for a qualitative answer, but ended up with the exact one! Well that certainly proves it beyond a shadow of a doubt.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

So, basically, solve the problem both ways. 

$$ V_{12} = M_{12} I_1 I_2 $$
$$ V_{21} = M_{21} I_1 I_2 $$

Let the toroid be 2 and the wire be 1.

The goal is to calculate the fluxes to achieve this.

$$ \phi_{12} = I_2 M_{12}$$ 

And vice-versa. Let's look at the toroid first. For a moment, we are rather confused about what exactly qualifies as the cross sectional area. If magnetic field lines came from above, would the toroid be a loop? But any circling magnetic field lines would flux through the coil itself.

Fortunately for us, wires only produce circling magnetic fields, so even if there is sometimes a case of ambiguous flux, here we can say for sure that there are *no* magnetic field lines to make that flux, so we consider the interior only to be that. If we model the torus as a series of current loops, we realize that each individual cross section has a flux roughly equal to that of **Problem 2!** flux! 

$$ \Phi_{12} = \mu_0 I_2 \left[ - \sqrt{d^2-R^2} + d \right] $$

Naturally we should exchange the variables for the ones relevant to this problem.

$$ \Phi_{21} = \mu_0 I_1 \left[ - \sqrt{(r_1+h/2)^2-(h/2)^2} + r_1+h/2 \right] $$

Where the numbers were swapped, R = h/2 as toruses are circular so the height is twice the radius. This means that the distance to the center is then $d = r_1+h/2$, as represented above. 

Given how the h/2 terms look  like they'll cancel, let's see how this simplifies. 

$$ \Phi_{21} = \mu_0 I_1 \left[ - \sqrt{r_1^2 + r_1h} + r_1+h/2 \right] $$

The temptation is to just multiply by N to get the flux along the entire torus, but no, we should probably integrate it. 

$$ \Phi_{total} = \int_0^{2\pi} \Phi_{21} n r d\theta $$

$$ \Rightarrow \Phi_{total} = 2\pi \Phi_{21} n r $$

Now n is the density of turns, which is $n = \frac{N}{2\pi r}$ in this case, which makes the result... $\Phi_{21} N$. So it looks like we could have just multiplied by N to begin with. Anyway, now divide by $I_1$. Notice that this cancels the I entirely. 

$$ M_{21} = \mu_0 N \left[ - \sqrt{r_1^2 + r_1h} + r_1+h/2 \right] $$

This matches the form of the problem found in the textbook on page 398, though naturally it is in a different form. So we can assert that this is definitely correct. Knowing this, we move on to the second verison. 

While we *thought* the magnetic field inside a torus was constant, turns out this is actually not the case. If we apply Ampere's law to a ring drawn through the torus, we note that we can draw rings of many different sized, rather than just one. $2\pi r B = \mu_0 N I$ or $B = \frac{\mu_0 N I}{2\pi r}$. This r is variable, but does not change from going in the vertical direction. (NI is the total current enclosed in this case). Note, however, that the magnetic field does not change vertically, so we can basically treat r as x, giving us this integral:

$$ \int_{r_1}^{r_1+h} \int_{-\sqrt{h^2/4-(x-r_1-h/2)^2}}^{\sqrt{h^2/4-(x-r_1-h/2)^2}} \frac{\mu_0 N I_2}{2\pi x} dydx $$

While it does not depend on y we still have to integrate over it. 

$$ = \int_{r_1}^{r_1+h} \frac{\mu_0 N I_2}{\pi x}\sqrt{h^2/4-(x-r_1-h/2)^2} dx $$

This integral is truly terrible. Simplifying it doesn't seem to go anywhere. The horror. So let's redefine the coordinate system, placing the center of the circle at the center.

$$ = \frac{\mu_0 N I_2}{\pi}\int_{-h/2}^{h/2} \frac{1}{x+r_1+h/2}\sqrt{h^2/4-x^2} dx $$

Which, while a bit ugly, is actually doable.



And thus began a rather slow drudgery of actually evaluating this at the bounds... except most of the arctan functions go to infinity and give neat results. 

$$ = \frac{\mu_0 N I_2}{\pi}[(r_1 + h/2)\pi - \sqrt{r_1}\sqrt{r_1+h}\pi]$$
$$ = \mu_0 N I_2[(r_1 + h/2) - \sqrt{r_1}\sqrt{r_1+h}]$$

This is the flux, we divide out the current for the inductance.

$$ M_{12} = \mu_0 N [(r_1 + h/2) - \sqrt{r_1}\sqrt{r_1+h}]$$

If we push the radicals together, we get...

$$ M_{12} = \mu_0 N [(r_1 + h/2) - \sqrt{r_1^2+r_1h}]$$

Which is EXACTLY the same as the previous, and we are done.
