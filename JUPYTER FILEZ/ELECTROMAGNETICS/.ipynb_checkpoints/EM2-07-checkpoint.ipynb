# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

NOTE: problem should read SIXTH circle, not THIRD. 

So, let's mathematically define our shape first. We have a cross section with radius R that goes from $\phi$ = 0 to $\frac{\pi}{3}$, and the waveguide itself extends in the z direction. We will clearly be using cylindrical coordinates for this. 

We can define TE and TM waves based on which of the longitudinal fields is zero. For TE, $E_z=0$, and for TM, $B_z = H_z = 0$. Note that H adn B do not in general equal each other we just have zero in that case. This logically makes sense, as the transverse part is the part that is "flat" with respect to the guide, and so the part that isn't transverse is what will have the z component. 

Anyway, the full equations in cylindrical coordinates for the longitudinal fields with $\gamma^2 = \mu \epsilon \omega^2 - h^2$ are...

$$ \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial E_z}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2 E_z}{\partial \phi^2} + \gamma^2 E_z = 0 $$
$$ \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial H_z}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2 H_z}{\partial \phi^2} + \gamma^2 H_z = 0 $$

We also have to consider the boundary conditions, which are the difference between E and H. E has to be 0 on teh gboundary, meanwhile the derivative of H across the boundary has to be zero. 

For now, let's just consider E. IN a circular waveguide, we assumed a solution of the form

$$ E_z = E(\rho) e^{im\phi} $$

Is this still reasonable in our case? In our case it is presumably better to use actual trig functions since we have hard limits on the edges here. We want it to be zero at zero for E, so make it a sine function. 

$$ E_z = E(\rho) sin(m\phi) $$

This also restricts values on m, as when $\phi =\frac{\pi}{3}$ the sine funciton needs to be zero, which means $\frac{m\pi}{3}$ must equal some $n\pi$, where n is an integer. 

Solving this gives us $m = 3n$, where n = 0, 1, 2, 3. So our solution is currently:

$$ E_z = E(\rho) sin(3n\phi) $$

Anyway, we now stick this back into the differential equation. 

$$ \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial  E(\rho) sin(3n\phi)}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2  E(\rho) sin(3n\phi)}{\partial \phi^2} + \gamma^2  E(\rho) sin(3n\phi) = 0 $$
$$\Rightarrow  sin(3n\phi) \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial  E(\rho)}{\partial \rho} \right) - \frac{1}{\rho^2} 9n^2 E(\rho) sin(3n\phi) + \gamma^2  E(\rho) sin(3n\phi) = 0 $$
$$\Rightarrow  \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial  E(\rho)}{\partial \rho} \right) - \frac{1}{\rho^2} 9n^2 E(\rho) + \gamma^2  E(\rho) = 0 $$
$$\Rightarrow \rho \frac{\partial}{\partial \rho} \left( \rho \frac{\partial  E(\rho)}{\partial \rho} \right) + (\gamma^2 \rho^2 - 9n^2 ) E(\rho) = 0 $$

Which is basically the same form as what we had for the circular case, the only difference being that m is 3n rather than just any integer. This does restrict our solutions rather impressively to just the bessel functions that are evenly divisible by 3. This does mean our solution is rather simple:

$$ E_z = E_0 J_{3n}(\gamma \rho) sin(3n\phi) $$

The result is very similar for H, except we use cosines since the *derivative* needs to be zero at the edges. This just makes sine a cosine. 

$$ H_z = H(\rho) cos(3n\phi)$$

Strangely enough this changes nothing for the derivation of the radial differential equation. Not even a sign difference.

$$\rho \frac{\partial}{\partial \rho} \left( \rho \frac{\partial  H(\rho)}{\partial \rho} \right) + (\gamma^2 \rho^2 - 9n^2 ) H(\rho) = 0 $$

$$ H_z = H_0 J_{3n}(\gamma \rho) cos(3n\phi)$$

These are the answers to part a)

$$ E_z = E_0 J_{3n}(\gamma \rho) sin(3n\phi) $$
$$ H_z = H_0 J_{3n}(\gamma \rho) cos(3n\phi)$$

Note that these only exist when the other is zero, as we are considering TE and TM only. 

Now what we're asked for part b) is what the frequency limit here is when $R = 3cm$. This means we get to apply the boundary conditions again, for at the edge E has to be zero and the derivative of H has to be zero, with respect to the normal vector of the surface which is $\hat \rho$ in both cases. Thus, we need...

$$ J_{3n}(\gamma R) = 0 $$
$$ J_{3n}'(\gamma R) = 0 $$

Naturally the lowest frequency will come from the lowest n (that isn't zero) so we now have...

$$ J_{3}(\gamma R) = 0 $$
$$ J_{3}'(\gamma R) = 0 $$

Now we all know we have to look these things up in tables, but we can still define $\gamma$ in terms of the zeroes we look up and R. In this case, $\gamma = zeroes/R$, for either the function itself or its derivative. So now we look up the properties of $J_3$...

The first zero for $J_3$ is 6.38016.

The first zero for $J_3$' is 4.20119.

Thus for TM, we have $\gamma = 6.38016/R$ and for TE we have $\gamma = 4.20119/R$

This identifies the TE case as the one that has the lower cutoff frequency, as $\gamma^2 = \mu \epsilon \omega^2 - h^2$. 

The final answer we seek, though, is $\omega$. Do we have everything we need to calculate this?

Keep in mind that technically we don't need every value to calculate, we just need to find the minimum value. In this case, we can make h very small, so small it vanishes to zero. Furthermore, we can assume the guide is empty, turning $\mu \epsilon$ into $\mu_0 \epsilon_0$, making it $1/c^2$. (speed of light in cm/s is 2.998e10) 

$$ \gamma^2 = \omega^2/c^2 \Rightarrow \omega = \gamma c $$.

Thus, actually calculating the lowest cutoff frequency gives us...

$$ \omega = (1.400 1/cm)(2.998e10 cm/s) = 4.198e10 1/s = 41.98 GHz $$

This corresponds to microwaves and wavelengths on the order of centimeters. Which makes sense as the opening itself is a few centimeters across.

Now, the Abstract itself:

Magnetic monopoles are a commonly theorized type of elementary particle that exhibits a magnetic charge akin to how electrons and protons exhibit an electric charge. If one allows for the existence of magnetic monopoles, Maxwell's Equations become completely symmetric and can be used to calculate virtually anything desired within the realm of electromagnetics. In this paper we explore the behavior of monopoles moving at relativistic speeds, and ask what kinds of fields would be required to confine such monopoles, with a focus on theoretically putting such monopoles in a particle accelerator. We accomplish this by taking the modified versions of Maxwell's equations and solving them with analytic techniques for the case of magnetic current rather than charge current, giving special care to relativistic effects such as synchrotron and bremsstrahlung radiation. From this we will devise a field that will confine the relativistic magnetic monopole in such a way that it can be readily used for experiments. Should analytic methods be insufficient, we will numerically integrate equations derived from Maxwell's using a custom differential equation solver. We expect to discover that the optimal setup will be almost exactly the same as modern particle accelerators, only that the confining magnetic field will be replaced with a confining electric field. Furthermore, as magnetic monopoles are expected to be truly massive, the strength of the fields will need to be scaled up significantly. Due to the larger mass, the monopole will have larger momentum, and so relativistic losses will likely increase, driving the power required up even further.
