# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

So, magnetic fields add. We already know the magnetic field of individual rings on the axis are. They are:

$$ \vec B_z(z) = \frac12 \mu_0 I \frac{R^2}{(z^2 + R^2)^{3/2}} \hat z$$

Now, if we move the ring up from the origin by b, we change this to...

$$ \vec B_z(z) = \frac12 \mu_0 I \frac{R^2}{((z-b)^2 + R^2)^{3/2}} \hat z$$

Which we would then add to an identical ring that is, nonetheless, on the other side of the origin and has a reverse contribution.

$$ \vec B_z(z) = \frac12 \mu_0 I \frac{R^2}{((z-b)^2 + R^2)^{3/2}} \hat z - \frac12 \mu_0 I \frac{R^2}{((z+b)^2 + R^2)^{3/2}} \hat z$$

Now, we care about the behavior around the origin. At z=0 this clearly goes to zero so that's not sufficient enough for us to work with. So we're going to need to find the first nonzero term here, what fun. Anyway, it may be best to combine some of the terms:

$$ \Rightarrow \vec B_z(z) = \frac12 \mu_0 I R^2 \left( \frac{1}{((z-b)^2 + R^2)^{3/2}} - \frac{1}{((z+b)^2 + R^2)^{3/2}} \right) \hat z $$

Now, since z is approaching zero we can eaisly have z<<R and z<<b both be true. Because of this, if we pull out $b^2$ from the denominators...

$$ \Rightarrow \vec B_z(z) = \frac12 \mu_0 I R^2 \left( \frac{1}{(b^2(z^2/b^2 - 2z/b + R^2/b^2 + 1))^{3/2}} - \frac{1}{(b^2(z^2/b^2 + 2z/b + R^2/b^2 + 1))^{3/2}} \right) \hat z $$

Now the first order approximation clearly goes to zero, so we keep the $z/b$ term as.

However, the $z^2/b^2$ term can go take a hike.

$$ \Rightarrow \vec B_z(z) = \frac{1}{2b^3} \mu_0 I R^2 \left( \frac{1}{( - 2z/b + R^2/b^2 + 1)^{3/2}} - \frac{1}{(2z/b + R^2/b^2 + 1)^{3/2}} \right) \hat z $$

Now we can expand this in terms of z/b, treating $R^2/b^2+1$ as a constant. Keeping only the first order terms, we ned up with...

$$ \Rightarrow \vec B_z(z) = \frac{1}{2b^3} \mu_0 I R^2 \left( \frac{1}{(R^2/b^2 + 1)^{3/2}} + \frac{3 z/b}{(R^2/b^2 + 1)^{5/2}}  - \frac{1}{(R^2/b^2 + 1)^{3/2}} + \frac{3 z/b}{(R^2/b^2 + 1)^{5/2}} \right) \hat z $$

$$ \Rightarrow \vec B_z(z) = \frac{1}{2b^3} \mu_0 I R^2 \frac{6 z/b}{(R^2/b^2 + 1)^{5/2}}  \hat z $$

$$ \Rightarrow \vec B_z(z) = \frac{3}{b^4} \mu_0 I R^2 \frac{z}{(R^2/b^2 + 1)^{5/2}}  \hat z $$

Which is clearly linear in z, which is all we sought to show.

Now, what we're being asked here is NOT to just find the value for b that makes the equaiton match, since we could make $B_0$ equal just about anything. What we're seeking is to minimize the higher order terms of the expansion, which means we need to go back and *not* ignore the higher order terms. 

$$ \vec B_z(z) = \frac{1}{2b^3} \mu_0 I R^2 \left( \frac{1}{(z^2/b^2 - 2z/b + R^2/b^2 + 1)^{3/2}} - \frac{1}{(z^2/b^2 + 2z/b + R^2/b^2 + 1)^{3/2}} \right) \hat z $$

We can still expand in z/b, but it will need to be a bit more precise to include the higher order terms. In order to keep things compact, let $R^2/b^2 + 1 = a$, where a is just some constant. Expanding both terms, we end up with the two following series: 

The first terms cancel, the third terms cancel, but the *fourth* term, the x^3 term, that won't cancel. We seek to minimize it with a choice of b. The full term would be:

$$ - \frac{1}{2b^3} \mu_0 I R^2 \frac{5(3a-7) z^3/b^3}{a^{9/2}} = - \frac{1}{2b^3} \mu_0 I R^2 \frac{5(3(R^2/b^2 + 1)-7) z^3/b^3}{(R^2/b^2 + 1)^{9/2}} $$

Now, minimizing this isn't exactly a simple endeavor. Ignoring the posibility of b=0, as that's a discontinuity, we find...



a graph that doesn't give us an even number value for b. This made us very suspicious... but apparently no, the right solution for a Helmholz coil is very different than that of a gradient coil. Which means... that a non-even number is still just fine! Thus, we just need to find the zero... which we can do, since there is a clear minus sign in the numerator. We just need to make that particular portion equal zero. 

$$ 3(R^2/b^2 + 1) - 7 = 0 \Rightarrow  3(R^2/b^2 + 1) = 7 $$
$$ \Rightarrow R^2/b^2 = 4/3 $$
$$ \Rightarrow 3R^2/4= b^2 $$
$$ \Rightarrow \sqrt{3/4}R = b $$

Which is roughly 0.87 if R=1, which matches our graph. We have found the zero, and even though it seems ugly, we remind ourselves that not every answer is simple or even has a clear pattern.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

Now one might imagine using superposition of rings to figure this out, but no, we're going to use the potential.

The potential does not funciton anywhere there is current. One would think this means we can't use it anywhere on the spherical shell, however, we do have some boundary conditions we can work with.

Thus we can immediately apply the spherical harmonics forms to the potentials, exterior and interior.

$$ \psi (r>R) = \sum_l A_l \left( \frac{R}{r} \right)^{l+1} P_l(cos\theta) $$
$$ \psi (r<R) = \sum_l B_l \left( \frac{r}{R} \right)^{l} P_l(cos\theta) $$

Which must equal each other at $\theta = 0$ and r=R. 

So before we continue we should probably figure out what, explicitly, the current density $\vec K$ is. We know the total charge, Q, and thus the surface charge density $\sigma = \frac{Q}{4\pi R^2}$. However at every spot on the sphere the tangential velocity, that is the *speed*, is going to be a bit different. 

Specifically, the tangential speed is $v = \omega R sin\theta$

Now, here's something curious. The units of current are amps, or charge over time, C/t. However, *surface* current would be C/mt, where m is a length in meters. Which $K = \sigma v$ is exactly right. Thus...

$$ \vec K = \frac{Q}{4\pi R} \omega sin\theta \hat \phi $$

Where we added the direction back in to be explicit. 

Now, we know the radial derivatives of the potential functions are identical at r=R. Which is to say...

$$ \sum_l -(l+1) A_l \left( \frac{R}{r} \right)^{l+1} \frac{1}{r} P_l(cos\theta) = \sum_l l B_l \left( \frac{r}{R} \right)^{l} 1/r P_l(cos\theta) $$

Now this is true at r=R, so make the substitution.

$$ \sum_l -(l+1) A_l \frac{1}{R} P_l(cos\theta) = \sum_l l B_l 1/R P_l(cos\theta) $$
$$ \Rightarrow -(l+1) A_l P_l(cos\theta) = l B_l P_l(cos\theta) $$
$$ \Rightarrow A_l = \frac{-l}{l+1} B_l $$

Now that we have $\vec K$ and our constants we can turn to our boundary conditions.

$$ \hat r \times (\vec B_o - \vec B_i) = \mu_0 \vec K $$

Now we *can* actually solve for the difference betwee the magnetic fields, though ti will be somewhat awkward.

$$ \hat r\times (\vec B_o - \vec B_i) = \frac{Q\mu_0}{4\pi R} \omega sin\theta \hat \phi $$

Now we are forced to contend with "inverting the cross product" which we need to be careful about. First, look up the cross product for radial coordinates. If we let our entire subtraciton be represetned by $\vec B^-$ we can represent the full cross product as the matrix

$$ \hat r \times \vec B^- = det \begin{bmatrix}
\hat r & \hat \theta & \hat \phi \\
1 & 0 & 0 \\
B^-_r & B^-_\theta & B^-_\phi
\end{bmatrix} $$

Now we originally thought there would be some finagling required to get the cross product to work on a sphere, but that turns out to only be a problem if the vectors we're working with are defined at different locations--in this case, our unit vectors are at a particlar point with repsect to the particlar magnetic field, so we're completely fine. 

So if we take the determinant, we end up with

$$ B^-_\phi \hat \theta - B^-_\theta \hat \phi = \frac{Q\mu_0}{4\pi R} \omega sin\theta \hat \phi $$

What does this tell us? Well, it tells us that $B^-_\phi$ has to be zero. This perhaps isn't a surprise, as the system we are working with is completley azimuthally symmetric, so there better not be any $\phi$ magnetic fields. We do have to ask what this means for the individual fields though, but from azimuthal symmetry we can also tell that there's zero dependence. 

However, what we get *no* information about is r-dependence, there could be anything in $B^-_r$ as the cross product utterly removes it. We'll have to rely on something else to find it. (It will pop out of the original form of the potentials).

One thing we *do* know, though, is that the portion that contributes to our boundary condition here is $B^-_\theta$. 

$$ -B^-_\theta = \frac{Q\mu_0}{4\pi R} \omega sin\theta $$
$$ \Rightarrow -(\vec B_{o\theta} - \vec B_{i\theta}) = \frac{Q\mu_0}{4\pi R} \omega sin\theta $$
$$ \Rightarrow -\frac{1}{r}\frac{d}{d\theta}(\psi_i - \psi_o) = \frac{Q\mu_0}{4\pi R} \omega sin\theta $$

Since this is the boundary condition, r=R. 

$$ \Rightarrow -\frac{1}{R}\frac{d}{d\theta}(\psi_i - \psi) = \frac{Q\mu_0}{4\pi R} \omega sin\theta $$
$$ \Rightarrow -\frac{d}{d\theta}(\psi_i - \psi_o) = \frac{Q\mu_0}{4\pi} \omega sin\theta $$

Now, let's just substitute everything in...

$$ \Rightarrow -\frac{d}{d\theta}(\sum_l B_l \left( \frac{r}{R} \right)^{l} P_l(cos\theta) - \sum_l \frac{-l}{l+1} B_l \left( \frac{R}{r} \right)^{l+1} P_l(cos\theta))  = \frac{Q\mu_0}{4\pi} \omega sin\theta $$

Once again, we are at r=R, since this is the boundary condition.

$$ \Rightarrow -\frac{d}{d\theta}\sum_l \left[ B_l P_l(cos\theta) - \frac{-l}{l+1} B_l P_l(cos\theta)\right]  = \frac{Q\mu_0}{4\pi} \omega sin\theta $$

$$ \Rightarrow -\frac{d}{d\theta}\sum_l \left[ \frac{1}{l+1} B_l P_l(cos\theta)\right]  = \frac{Q\mu_0}{4\pi} \omega sin\theta $$

The derivative can move into the sum.

$$ \Rightarrow -\sum_l \left[ \frac{1}{l+1} B_l \frac{d}{d\theta} P_l(cos\theta)\right]  = \frac{Q\mu_0}{4\pi} \omega sin\theta $$

Now we could take the derivative with the chain rule, but we can recognize something here: the final form our answer needs to take needs to be purely $sin\theta$ in nature. Since the Legandre polynomials are orthogonal (and their derivatives would be too) if one of the polynomials matches the condition, all of them have to. And that just so happens to be $P_1$

$$ P_1(cos\theta) = cos\theta ; P'_1(cos\theta) = -sin\theta $$

This means **l must equal 1**. 

$$ \Rightarrow \frac{1}{2} B_1 \sin\theta  = \frac{Q\mu_0}{4\pi} \omega sin\theta $$

$$ \Rightarrow B_1  = \frac{Q\mu_0}{2\pi} \omega $$

All other B values are zero. This gives us full potentials of:

$$ \psi (r>R) = \frac{-1}{2} \frac{Q\mu_0}{2\pi} \left( \frac{R}{r} \right)^{2} cos\theta $$
$$ \psi (r<R) = \frac{Q\mu_0}{2\pi} \omega \frac{r}{R} cos\theta $$

The magnetic field is rather simple to find with $\psi$

$$ \vec B = - \nabla \psi $$

Now as our funcitons only depend on r and $\theta$ we don't have to worry about basically anything else. Just take the derivatives. 

$$ \vec B_o = - \frac{Q\mu_0\omega}{2\pi} \frac{R^2}{r^3} cos\theta \hat r - \frac{1}{2} \frac{Q\mu_0\omega}{2\pi} \frac{R^2}{r^2}  sin\theta \hat \theta $$

$$ \vec B_i = - \frac{Q\mu_0\omega}{2\pi} \frac{1}{R} cos\theta \hat r + \frac{Q\mu_0\omega}{2\pi} \frac{r}{R} sin\theta \hat \theta $$

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

We can consider this as two loops. The moments in each loop can be thought of as m = IA with direction given by the right hand rule. Each loop has current I, and area $a^2$. 

So the bottom loop is $\vec m = - Ia^2 \hat z$ and the top loop is $\vec m = + Ia^2 \hat z$

The only problem here is that these moments can't just add since they are at different *locations.* So let's figure out how exactly to use this information.



Now, we are being asked for A, the vector potential. This is, in simple form, given by:

$$ \vec A = \frac{\mu_0}{4\pi} \frac{\vec m \times \vec r}{r^3} $$

Naturally different As add, we just need to watch the offset. In our case, this turns into:

$$ \vec A = \frac{\mu_0}{4\pi} Ia^2 \left( \frac{\hat z \times \vec r_1}{r_1^3} - \frac{\hat z \times \vec r_2}{r_2^3} \right) $$

Where

$$ \vec r_1 = \vec r + a/2 \hat y ; \vec r_1 = \vec r - a/2 \hat y  $$

Now the part that usually makes this easy is that the offset is in the z direction and it vanishes in the cross product. Not so today. Instead, we get this "mess:"

$$ \vec A = \frac{\mu_0}{4\pi} Ia^2 \left( \frac{\hat z \times \vec r}{|\vec r + a/2 \hat y|^3} + \frac{\hat z \times a/2 \hat y}{|\vec r + a/2 \hat y|^3} - \frac{\hat z \times \vec r_2}{|\vec r - a/2 \hat y|^3} + \frac{\hat z \times a/2 \hat y}{|\vec r - a/2 \hat y|^3} \right) $$

So, first, let's examinet he denominators. Using $y = rsin\theta sin\phi$ we can convert them into more manageable forms. All of them become $(r^2 + a^2/4 \pm rasin\theta sin\phi)^{3/2} \approx r^3(1 \pm (a/r)sin\theta sin\phi)^{3/2}$

This allows us to "simplify" to

$$ \vec A = \frac{\mu_0 I a^2}{4\pi r^3} \left( \frac{\hat z \times \vec r}{(1 + (a/r)sin\theta sin\phi)^{3/2}} + \frac{\hat z \times a/2 \hat y}{(1 + (a/r)sin\theta sin\phi)^{3/2}} - \frac{\hat z \times \vec r_2}{(1 - (a/r)sin\theta sin\phi)^{3/2}} + \frac{\hat z \times a/2 \hat y}{(1 - (a/r)sin\theta sin\phi)^{3/2}} \right) $$

At this point it's pretty evident we should perform the first order expansion. Since we are definitely looking for the asymptotic form, we assume large r thus a >> r and expand around a/r. The result of the expansiosn (considering only denominators) is $1 \pm \frac32 (a/r) sin\theta sin\phi + O(a^2/r^2) $.



For the z cross r terms, the constant term cancels. However, for the z cross y terms, the linear term cancels! Thus, we can look at the two results and see:

$$ \vec A = \frac{\mu_0 I a^2}{4\pi r^3} \left( -(\hat z \times \vec r)\frac32 \frac{a}{r} sin\theta sin\phi + 2(\hat z \times a/2 \hat y) \right) $$

So we can actually evaluate the z cross y product rather reasily.

$$ \vec A = \frac{\mu_0 I a^2}{4\pi r^3} \left( -(\hat z \times \vec r)\frac32 \frac{a}{r} sin\theta sin\phi - a \hat x \right) $$

And the cross product between z and r is also known.

$$ \vec A = \frac{\mu_0 I a^2}{4\pi r^3} \left( - \frac{3a}{2} sin^2\theta sin\phi \hat \phi - a \hat x \right) $$

Now, one could say this is an answer, although the unit vectors do not agree with each other. We *are* asked to find it in cartesian, so we need to start making some conversions. We just have to dig around a bit to find unit vector conversion for $\hat \phi$, whcih is $\hat \phi = cos\phi \hat x + sin\phi \hat y $

$$ \vec A = \frac{\mu_0 I a^2}{4\pi r^3} \left( - \frac{3a}{2} sin^2\theta sin\phi cos\phi \hat x - \frac{3a}{2} sin^2\theta sin^2\phi \hat y - a \hat x \right) $$

Now we need to get rid of the angles. In this case, $x = rsin\theta cos\phi; y = rsin\theta sin\phi$ When we make these substitutions a further $1/r^2$ pops out since the r's are not present in the terms.

$$ \vec A = \frac{\mu_0 I a^2}{4\pi r^5} \left( - \frac{3a}{2} xy \hat x - \frac{3a}{2} y^2 \hat y - a \hat x \right) $$
$$ = \frac{\mu_0 I a^2}{4\pi r^5} \left( - (\frac{3a}{2} xy  + a)\hat x - \frac{3a}{2} y^2 \hat y\right) $$
$$ = \frac{\mu_0 I a^3}{4\pi r^5} \left( - (\frac{3}{2} xy  + 1)\hat x - \frac{3}{2} y^2 \hat y\right) $$

Anyway, the z-dependence is hiding within r. We like it stated this way, since it's clear what things depend on what and where. However, we are told to use all cartesian, so we might as well write out r for our final answer.

$$ \vec A = \frac{\mu_0 I a^3}{4\pi (x^2+y^2+z^2)^{5/2}} \left( - (\frac{3}{2} xy  + 1)\hat x - \frac{3}{2} y^2 \hat y\right) $$

Which is the behavior of the A-field (the vector potential) far from the "8"-shaped system itself. Notably there is nowhere it points in the z direction, which mathematically makes sense, all the direcitons come from a cross product with the z unit vector, so all of the z-directions have to cancel out. We have *cycling* occur here. (naturally the magnetic field itself will have some z-components as we can relate A to B via the curl, which will involve a z-derivative). 

The most interesting point is what happens at x=0 and y=0, which is to say directly above the 8 in the center, between the two moments (but far away.)

$$ \vec A = \frac{\mu_0 I a^3}{4\pi z^5} \left( -\hat x \right) $$

Now we note that the two loops of current interfere *constructively* in the center, so it makes sense that the potential in the middle would be in that direction. We note that the laplacian of A is related to the *negative* of the current density, which explains the sign being negative. 

Anyway, with this, we consider the problem complete.
