# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

If it's polarized in parallel, then $\vec P = P\hat z$, where P is just the magnitude of polarization. 

We note that this polarization will end up, at a distance, taking the form of a dipole or something similar as the polarization makes there be a positive charge buildup on one end and a negative on the other. 



Originally, we used Poisson's Formula, but it was pointed out that was overkill since we know the polarization we can just find the bound charge, that is, the charge that influences the created field. 

$$\rho_{net} = -\nabla \cdot \vec P$$

Only the z component exists, so we end up with:

$$\rho_{net} = 0 $$

However, this is just the volume charge density, we do have to concern ourselves with the charge at the ends. 

$$\sigma_b = \vec P \cdot \hat n$$

Now the unit vector is just +z or -z at the ends of the cylinder, which is effectively just a plus or minus sign applies to the polarization. 

$$\sigma_{btop} = P$$
$$\sigma_{bbottom} = -P$$

Which means we have a positively charged disc on top and a negatively charged disc on the bottom. The rest of the dielectric effectively has no charge. 



Now, what of the two discs? Well, we have the disc from example 3.1

$$ \vec E_{disc}(z>0) = \frac{\sigma}{2\epsilon_0} \left[ 1-\frac{z}{\sqrt{z^2+R^2}} \right] \hat z $$
$$ \vec E_{disc}(z>0) = \frac{\pm P}{2\epsilon_0} \left[ 1-\frac{z}{\sqrt{z^2+R^2}} \right] \hat z $$

We need to adjust this to be for discs placed at L/2 and -L/2, which will have their own effects. Each "z" refers to how far "above" the disc they are, so in the case of the lower disc it's L/2+z, and in the case of the upper disc it's L/2-z, but in the upper disc the push is in the reverse direction so it will point opposite--but the sign of P is also different.. Therefore...

$$ \vec E_{total} = -\frac{P}{2\epsilon_0} \left[ 1-\frac{L/2-z}{\sqrt{(L/2-z)^2+R^2}} \right] \hat z + \frac{ - P}{2\epsilon_0} \left[ 1-\frac{(L/2+z)}{\sqrt{(L/2+z)^2+R^2}} \right] \hat z $$

When z=0, everything will combine...

$$ \boxed{\vec E_{total} = -\frac{P}{2\epsilon_0} \left[ 2-\frac{L}{\sqrt{(L/2)^2+R^2}} \right] \hat z} $$

For the record, this is the exact same solution we got as with Poisson's Formula, so we are highly confident in the result. 


Notably this does NOT have a uniform polarization, so we can't take the shortcut of Poisson's formula. Instead we turn to the more general 6.13...

$$ \vec E_p (\vec r) = \frac{1}{4\pi\epsilon_0} \int_V d^3r' \frac{-\nabla' \cdot \vec P(\vec r')}{|\vec r - \vec r'|^3}(\vec r - \vec r') + \frac{1}{4\pi\epsilon_0} \int_S d\vec S' \cdot \vec P(\vec r') \frac{\vec r - \vec r'}{|\vec r - \vec r'|^3} $$

Which is, quite obviously, too complicated. In fact we overcomplicated part a, but we already have an answer for that.

What we need to do is actually use the polarization to find the bound charge distribution.

$$\rho_{net} = -\nabla \cdot \vec P$$

Only the z component exists, so we end up with:

$$\rho_{net} = -\frac{2P_0}{L} $$

This is the bound charge distribution. It depends on no variables whatsoever. Helpful. Naturally this is only true inside the cylinder, outside there has to be no charge. 

Now that is the charge density in the bulk of the material, we also know there is some positive and negative things on the top and bottom of the cylinder at -L/2 and L/2. Fortunately the boundary conditions here are $\sigma_b = \vec P \cdot \hat n$, where the n normal vector is just z-up or z-down depending. on which edge we're at. Thus...

$$\sigma_{btop} = \frac{2P_0z}{L} = P_0$$
$$\sigma_{bbottom} = -\frac{2P_0z}{L} = P_0$$

So ultimately we can visualize this cylinder as actually having a polarization where the positive charge is "stretched" out so it builds up on the ends of the cylinder. Neat. Anyway this means we can consider the problem a cylinder of constant chage density and two discs of constant charge density. 

We have the interior of a cyilnder states in terms of 0 to L.

$$ \textbf{E}(z<L) = \frac{\rho}{2\epsilon_0} \left[ 2z-L -\sqrt{z^2+R^2} +\sqrt{(L-z)^2+R^2} \right] \hat z $$

But we need it in the form of -L/2 to L/2. This is, fortunately, the form given in Dr. Hedman's answers for the problem in homework 2. 

$$ \textbf{E}(z<L) = \frac{\rho}{2\epsilon_0} \left[ 2z -\sqrt{(z+L/2)^2+R^2} +\sqrt{(z-L/2)^2+R^2} \right] \hat z $$

Now we have $\rho$ as the volume charge right here, so we insert it and get...

$$ \vec E_{volume} = -\frac{P_0}{L\epsilon_0} \left[ 2z -\sqrt{(z+L/2)^2+R^2} +\sqrt{(z-L/2)^2+R^2} \right] \hat z $$

Now, what of the two discs? Well, we have the disc from example 3.1

$$ \vec E_{disc}(z>0) = \frac{\sigma}{2\epsilon_0} \left[ 1-\frac{z}{\sqrt{z^2+R^2}} \right] \hat z $$
$$ \vec E_{disc}(z>0) = \frac{P_0}{2\epsilon_0} \left[ 1-\frac{z}{\sqrt{z^2+R^2}} \right] \hat z $$

We need to adjust this to be for discs placed at L/2 and -L/2, which will have their own effects. Each "z" refers to how far "above" the disc they are, so in the case of the lower disc it's L/2+z, and in the case of the upper disc it's L/2-z, but in the upper disc the push is in the reverse direction so it will point opposite. Therefore...

$$\vec E_{total} = -\frac{P_0}{L\epsilon_0} \left[ 2z -\sqrt{(z+L/2)^2+R^2} +\sqrt{(z-L/2)^2+R^2} \right] \hat z + \frac{P_0}{2\epsilon_0} \left[ 1-\frac{L/2+z}{\sqrt{(L/2+z)^2+R^2}} \right] \hat z - \frac{P_0}{2\epsilon_0} \left[ 1-\frac{L/2-z}{\sqrt{(L/2-z)^2+R^2}} \right] \hat z $$

Technically this is an answer, but we can combine some terms. For instance, a fair chunk of the disc sections completley cancel. 

$$\vec E_{total} = -\frac{P_0}{L\epsilon_0} \left[ 2z -\sqrt{(z+L/2)^2+R^2} +\sqrt{(z-L/2)^2+R^2} \right] \hat z + \frac{P_0}{2\epsilon_0} \left[-\frac{2z}{\sqrt{(L/2+z)^2+R^2}} \right] \hat z $$

And there's not really any obvious terms to simplify at this point. However, we are asked to check the value at the center, which would be when z=0. When z=0, though, the entire field goes to zero! 

This is to be expected. Remember, we calculated the polarization to create a cylinder with a solid negative charge distribution, which at the center would cancel out, and two positive caps, which at equal distances from the origin would cancel out. Thus, it all vanishes at the center. Unlike the previous part, which is a uniform polarization.

We note that this also makes physical sense--the "arrows" of hte polarizaiton point opposite directions as they go away from the center of the cylinder, so naturally it all cancels out in the center.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

So, let's consider the "start" situation first. That is, the dielectric is not under a field and is thus neutral, and a point charge is added. So our "initial" electric field is just that of a point charge.

$$ \vec E_i = \frac{Q}{4\pi\epsilon_0 r^2} \hat r $$

Now we must of course be careful not to make the same mistake we were warned against: this is not the total electric field, for once the shell is exposed to an electric field, it's going to produce its own electric field in response. The assumption we make is that the solution will be very similar to that of a conductor in the same situation, that is, a shell subject to the field from a point charge. Thus, we should label are areas.

Area I is the area inside the shell that is concerned only with the point charge by Gauss' law. Area II is the area inside the shell itself--for the conductor, this would have a uniform electric field. Area III is outside the shell.

Area I has no change between conductor and dielectric situations. The electric field is just $\frac{Q}{4\pi\epsilon_0 r^2} \hat r$ and the potential is $\frac{Q}{4\pi\epsilon_0 r} \hat r$

Area II, in the conudcting case, has to have an electric field of zero (and a constant potential). This means for a conductor the situation of the charge density is rather simple: -Q on the inside of the sphere to cancel out the point charge, and then +Q on the outside to balance out conservation of charge. The density would be determined by the surface area. And area III would look like a single point charge from a distance--the conductor would essentially be invisible.

This last factoid has to be true of the dielectric as well. Its overall charge is neutral, so from far away it must look invisible. This allows us to define some things rather rigidly--for instance, the electric field in area III is the same as one point charge, or $\frac{Q}{4\pi\epsilon_0 r^2} \hat r$. However, the potential can be quite different, for the electric field inside the dielectric need not be exactly zero, nor does it need to cancel out the electric field from the source charge. 

So, if we examine the contribution to the electric field of the conductor version alone, we see that it has to cancel out the point charge's field. So the conductor alone, inside, produces a field of $ -\frac{Q}{4\pi \epsilon_0 r^2}\hat r $. Essentially, the conductor perfectly cancels out the field, giving the total zero.

Now the dielectric isn't going to be this effective, it's going to be some constant times that. Let's call it A. Since the dielectric is spherically symmetric and electrically neutral, it can't influence regions I and III, only II. 

$$ \vec E_{dII} = -A\frac{Q}{4\pi\epsilon_0r^2} \hat r $$

And the other source is the point charge, which is the same in all regions, $\vec E_{p} = \frac{Q}{4\pi \epsilon_0 r^2}$



Notably the dielectric provides what appears to be the field of a negative point charge at the center, just one not as "equal" to the original point charge Q. This makes sense considering we expect to find a shell of negative charge hanging out on the inner surface, Gauss law expects to treat that as its own negative point charge. 

One might start going to potentials at this point... but we just realized, we have the entire electric field. 

$$ E_{I} = E_{III} = \frac{Q}{4\pi\epsilon_0 r^2} $$
$$ E_{II} = \frac{Q}{4\pi\epsilon_0 r^2} - A\frac{Q}{4\pi\epsilon_0 r^2} $$

Where A is, presumably, less than 1. What is it, though? Well, whatever it is, it needs to make the potentials constant. The problem is that potential is a little *odd* for situations like this. See, when we take the derivatives...

$$ V_{I} = V_{III} = \frac{Q}{4\pi\epsilon_0 r} + C $$
$$ V_{II} = \frac{Q}{4\pi\epsilon_0 r} - A\frac{Q}{4\pi\epsilon_0 r} + D $$

Where C and D are constants of integration, and annoying ones at that. We can remove them by declaring V=0 at infinity, but since we then have to come in *from* infinity, that means area I and III are actually not the same in regards to potential, while they are in electric field.

$$ V_{III} = \frac{Q}{4\pi\epsilon_0 r} $$
$$ V_{II} = \frac{Q}{4\pi\epsilon_0 r} - A\frac{Q}{4\pi\epsilon_0 r} + A\frac{Q}{4\pi\epsilon_0 R_2} $$
$$ V_{I} = \frac{Q}{4\pi\epsilon_0 r} - A\frac{Q}{4\pi\epsilon_0 R_1} + A\frac{Q}{4\pi\epsilon_0 R_2} $$

Which means we can't just up and set anything equal at a boundary to get A, since that just cancels A out every time on the boundary. We need more information. 

Let's try the D-field. It should also be continuous along each of the boundaries. In places with no dielectric, $\vec D = \epsilon_0 \vec E$, and with a dielectric $\vec D = \epsilon \vec E$ so we can write out our D-fields...

$$ D_{III} = D_{I} = \epsilon_0 \frac{Q}{4\pi\epsilon_0 r^2} $$
$$ D_{II} = \epsilon\left( \frac{Q}{4\pi\epsilon_0 r^2} - A\frac{Q}{4\pi\epsilon_0 r^2} \right) = \epsilon_0 (1+\frac{r}{R_x})\left( \frac{Q}{4\pi\epsilon_0 r^2} - A\frac{Q}{4\pi\epsilon_0 r^2} \right)  $$

We will be treating $R_x$ as just some constant. 

Now, when r=$R_2$, what does A have to equal to maintain D continuity?

$$ \epsilon_0 \frac{Q}{4\pi\epsilon_0 R_2^2} = \epsilon_0 (1+\frac{R_2}{R_x})\left( \frac{Q}{4\pi\epsilon_0 R_2^2} - A\frac{Q}{4\pi\epsilon_0 R_2^2} \right) $$

$$ \Rightarrow 1 = (1+\frac{R_2}{R_x})( 1 - A ) $$
$$ \Rightarrow 1 = 1 + \frac{R_2}{R_x} - A - \frac{R_2A}{R_x} $$
$$ \Rightarrow 0 = \frac{R_2}{R_x} - A(1 + \frac{R_2}{R_x}) $$
$$ \Rightarrow \frac{R_2}{R_x} = A(1 + \frac{R_2}{R_x}) $$
$$ \Rightarrow \frac{\frac{R_2}{R_x}}{1 + \frac{R_2}{R_x}} = A $$

Now if $R_x$ was equal to, say, $R_2$ this would just reduce to 1/2, but we'll leave it as is. Notably this is always going to be less than 1 (barring limit situations), which is what we expected A to be. This gives us an electric field inside the dielectric as

$$ E_{II} = \frac{Q}{4\pi\epsilon_0 r^2} - \frac{\frac{R_2}{R_x}}{1 + \frac{R_2}{R_x}}\frac{Q}{4\pi\epsilon_0 r^2} $$

Now Polarization can be achieved by $P = (\epsilon - \epsilon_0) E$.

$$ P = \frac{\epsilon_0 r}{R_x} \frac{Q}{4\pi \epsilon_0 r^2} (1 - \frac{\frac{R_2}{R_x}}{1 + \frac{R_2}{R_x}}) $$
$$ P = \frac{Q}{4\pi rR_x} (1 - \frac{\frac{R_2}{R_x}}{1 + \frac{R_2}{R_x}}) $$

$$\rho_{net} = -\nabla \cdot \vec P$$

Now hold it! We've been operating rather loosely, but at this point we're not just converting between potential and field, but we are in spherical coordiantes, so we need to be careful! What is the spherical divergence for something with just an r dependency?

$$ \nabla \cdot \vec A = \frac{1}{r^2} \frac{\partial}{\partial r} (r^2 A_r) $$

Which means our volume bound charge ends up being...

$$\rho_{net} = -\frac{Q}{4\pi R_x}(1-\frac{\frac{R_2}{R_x}}{1 + \frac{R_2}{R_x}})\frac{1}{r^2} $$

Which essentially says negative charge is distributed in the middle of the shell, more densely at the center than at the edge. 

Now the edge situation is 

$$\sigma_b = \vec P \cdot \hat n$$

Fortunately the unit vector and the polarization point the same direction, as we are radially symmetric, so we don't need to be concerned about funky coordinate issues like we were above. The only thing we are concerned with is the sign of the unit vector: positive on the outer surface, negative on the interior surface. 

$$\sigma_{bout} = \frac{Q}{4\pi R_2R_x} (1 - \frac{\frac{R_2}{R_x}}{1 + \frac{R_2}{R_x}})$$
$$\sigma_{bin} = -\frac{Q}{4\pi R_1R_x} (1 - \frac{\frac{R_2}{R_x}}{1 + \frac{R_2}{R_x}})$$

Which we note has the predicted behavior. Positive buildup on the outer edge, negative on the inner edge. The density is notably higher for the inner edge, but this is because there's less surface for the charge to go to, so this is still sensible.

It occurs to us that our original answer was meandering--a more technical and illuminating way to do this would be to show that the arrangement of charges forms a dipole, we know dipolar fields are not purely radial. We note that the only charges are surface charges (see part ii below) and so the field is created entirely by the dipolar shells. We do not need to know exactly what these charges are, we just need to prove that they're dipolar in arrangement. $\sigma_b = \vec P \cdot \hat n$ is the surface charge distribution relation. The problem is, we can't find the polarizaiton directly, for $\vec P = (\epsilon - \epsilon_0)\vec E$, and E is part of the thing we're trying to find. However...

$$\epsilon - \epsilon_0 = \frac{\epsilon_0}{1-\alpha cos\theta} - \epsilon_0 = \frac{\alpha cos\theta}{1-\alpha cos\theta}$$

Which, by itself, could be used to show that E cannot be purely radial since no solution exists with purely radial E for this weird setup, but we want to show that it is a pure dipole for later parts. So, let's consider the D-field. One thing we can know about the D-field is that it is related to the *free charge*, and the free charge is all the charge that's "free to move" (hence the name) which means it responds equally in all directions, thus D must be spherically symmetric (i.e., purely radial). This means the central D field will depend only on the central point charge! That is, 

$$ D = \frac{Q}{4\pi r^2} \hat r$$

Which is derived from the fact that, in the absence of a dielectric, $D = \epsilon_0 E$, so we grab the free electric field and throw it in. However, we do have a dielectric, so $D=\epsilon E$ can be used to find E! 

$$ E = \frac{Q}{4\pi \epsilon_0 r^2} (1-\alpha cos\theta) \hat r$$

Which obviously has an anguylar component. From $P = (\epsilon - \epsilon_0) E$ we get

$$ P = \frac{Q}{4\pi r^2} \alpha cos\theta $$

Which, yes, is the dipole. So we are now convinced that the potential is the dipole potential, which definitely varies the electric field with $\theta$. But there's one odd thing here--isn't the E the answer we were looking for?

The answer is no, this is just ONLY the field from the central charge acting here, we haven't factored in the field from the dielectric acting on itself. However, the dielectric would be of the same form as this, so the overall behavior would still be that of a dipole. The exact strength of that dipole, well, that's what part iii is form.

In conclusion: even if this is all somewhat shaky, E most definitely depends on $\theta$. That said, it appears to only have a radial component, no sideways vectors. Which makes sense, considering that a point charge is driving the field, ultimately. 

Holding the angle constant shows us that the dielectric constant is just that, a *constant*, as we move along a radius r. There's no variation from one level of r to the other, variation only occurs at the boundaries if we move straight out. This means the polarization will respond *directly* to the strength of E everywhere, meaning that all the charges get distributed around evenly without one part being adjusted more than any other, and everything gets pushed perfectly.

Mathematically, $\rho_b = -\nabla \cdot \vec P$. When we consider only the radial direction of P (which is what we're doing in this illustration as that's the place between the two boudnaries), the polarization will take the form of the electric field, something relating to $\frac{1}{r^2}$. When taking the divergence in radial coordinates, this goes to 0. It just up and vanishes. So the volume charge does not exist. 

So, we are making the assumption that our result in part i is correct, and that the overall field the dielectric produces will be that of a dipole. We also note that ii is correct and that there is no volume charge whatsoever. 

The curious thing is that both of the surface charges are dipoles, and they will have to cancel each other out to achieve electric neutrality. (Though, admittedly, this is only a cancelation on large scales, near the surfaces the field wont' seem exactly like what we expect.) The interiors of dipoles produce uniform electric fields, while the exteriors have the dipole potential. 

To start with, all regions have the point charge potential $ \frac{Q}{4\pi\epsilon_0 r}$, this is the same in all of them no matter what. 

The center region is inside both dipoles, which means both the inner and outer shell provide a potential of Ar, though they will have different A values. But we can combine them together so...

$$\varphi_{I} = \frac{Q}{4\pi\epsilon_0 r} + A_1r $$

Now inside the dielectric, the only position that changes is that of the inner surface. This switches its potential to external dipole...

$$\varphi_{II} = \frac{Q}{4\pi\epsilon_0 r} + B_1\frac{1}{r^2}cos\theta + A_2r $$

Then outside everything, all of it combines into a single dipole term...

$$\varphi_{III} = \frac{Q}{4\pi\epsilon_0 r} + B_2\frac{1}{r^2}cos\theta $$

Now, at r=$R_1$, these first two have to be the same. The influence of the central charge does not change over the boundary, so we can essentially ignore it. 

$$ A_1R_1 = B_1\frac{1}{R_1^2}cos\theta + A_2R_1 \Rightarrow A_1 - A_2 = B_1\frac{1}{R_1^3}cos\theta $$

Something smilar arises at the $R_2$ boundary. 

$$ B_1\frac{1}{R_2^2}cos\theta + A_2R_2 = B_2\frac{1}{R_2^2}cos\theta \Rightarrow A_2 = (B_2-B_1)\frac{1}{R_2^3}cos\theta$$

Sadly we have two equations and four unknowns here... but we might be able to pull a trick since we only need to solve, exactly for $A_1$. Furthermore we cna use the D-field here, since it better be continuous across the dielectric at a place where there is no built up charge, which happens at $\theta = \pi/2$, which is the equator of the dipole region. Here, $\epsilon_0 E_{I} = \epsilon E_{II}$ and $\epsilon_0 E_{III} = \epsilon E_{II}$, on the boundaries anyway. 

So to relate this properly, we can extract it from our potential relations above. Note that the point charge potentials do not cancel due to the multiplication. 

$$ \epsilon_0 A_1 + \epsilon_0 \frac{Q}{4\pi\epsilon_0 R_1} = \epsilon \frac{Q}{4\pi\epsilon_0 R_1} -\epsilon B_1 \frac{2}{R^3_1} cos\theta + \epsilon A_2$$
$$ \epsilon_0 \frac{Q}{4\pi\epsilon_0 R_2} -\epsilon_0 B_2 \frac{2}{R^3_2} cos\theta = \epsilon \frac{Q}{4\pi\epsilon_0 R_2} -\epsilon B_1 \frac{2}{R^3_2} cos\theta + \epsilon A_2 $$

The negative signs show up due to taking derivatives of the dipolare terms. Now we set the angle to $\pi/2$

$$ \epsilon_0 A_1 + \epsilon_0 \frac{Q}{4\pi\epsilon_0 R_1} = \epsilon \frac{Q}{4\pi\epsilon_0 R_1} -\epsilon B_1 \frac{2}{R^3_1} + \epsilon A_2$$
$$ \epsilon_0 \frac{Q}{4\pi\epsilon_0 R_2} -\epsilon_0 B_2 \frac{2}{R^3_2} = \epsilon \frac{Q}{4\pi\epsilon_0 R_2} -\epsilon B_1 \frac{2}{R^3_2} + \epsilon A_2 $$

And now we have four equations, four unknowns. This should be doable. We DO set their angles, since all equations have to be true for all $\theta$ anyway. 

$$ A_1 - A_2 = B_1\frac{1}{R_1^3}$$

$$ A_2 = (B_2-B_1)\frac{1}{R_2^3}$$

Now we just do linear algebra. Since it's decidedly not simple linear algebra, we set the constants we were interested in to xyzw and made the rest of the terms special constants. Also because we were getting a little tired of this problem. And wouldn't you know it, even in absurd generalities it spit out an answer. 



So at least the answer exists. That said, we're going to have to figure it out the hard way since simplifying that won't go easy. The best way to do this appears to be to set up a matrix and see if there's any patterns we can work out. 

$$\begin{bmatrix}
\epsilon_0 & -\epsilon & \epsilon\frac{2}{R^3_1} & 0 \\
0 & -\epsilon & \epsilon\frac{2}{R^3_2} & -\epsilon_0\frac{2}{R_2^3} \\
1 & -1 & -\frac{1}{R^3_1} & 0 \\
0 & 1 & \frac{1}{R^3_2} & -\frac{1}{R_2^3} 
\end{bmatrix}\begin{bmatrix}
A_1 \\
A_2 \\
B_1 \\
B_2 
\end{bmatrix}=\begin{bmatrix}
(\epsilon-\epsilon_0) \frac{Q}{4\pi\epsilon_0 R_1} \\
(\epsilon-\epsilon_0) \frac{Q}{4\pi\epsilon_0 R_2} \\ 
0  \\
0 
\end{bmatrix}$$

Ultimately the only one we need to solve for is $A_1$, so if we can find a trick to do that, we're golden. After much trial and error, we found a rather complciated method. Rather than typing up every step (we'd be here for pages) we'll outline it. First, the $A_2$ variable is in a clear space where it can be canceled out repeatedly in multiple ways, so taking vairous combinations we created equations only in terms of $A_1, B_1,B_2$, and we ended up with four entirely independent equations this way. (or well, it looks like they're independent, naturally there must be some crossover as they do dictate a single solution for $A_2$. From here, we solved for $A_1$ and got four $A_1 = f(B_1,B_2)$ equations. We then set these equal to each other to get relations of $B_1$ and $B_2$ in terms of each other only. We solved for $B_1$ in terms of $B_2$, substituted, and arrived at long last for a value for $B_2$

$$ B_2 = \frac{-\epsilon(1+\epsilon)QR_2^2}{8\pi \epsilon_0} $$

Which let us quickly solve for $B_1$

$$ B_1 = \frac{\epsilon Q R_2^2}{8 \pi \epsilon_0} \left[ -\frac13 \epsilon - \frac23 \epsilon_0 - 1 \right] $$

Since we now have both of these, we can use one of our many other equations to find $A_1$. After some re-arranging, we get...

$$ A_1 = \frac{\epsilon Q R_2^2}{8 \pi \epsilon_0} \left[  \frac{2\epsilon_0}{3R_2^3} - \frac{2\epsilon}{3R_2^3} - \frac{1}{R_2^3} - \frac{2\epsilon_0}{3R_1^3} - \frac{\epsilon}{3R_1^3} \right] $$

And now we COULD solve for $A_2$... but why bother? We only need $A_1$ for the area we found. We note that there is probably a simpler way to write out $A_1$ but its current form is usable. It also somewhat sensibly depends both on $R_1$ and $R_2$ since we feel both fields. 

As a reminder, the potential is:

$$\varphi_{I} = \frac{Q}{4\pi\epsilon_0 r} + A_1r $$

Now, the electric field is $E = -\nabla V$. We can just take the derivative, right? Well... no. See, $A-1$ may look like it doesn't depend on $\theta$, but $\epsilon$ does, meaning when we take the derivative their can and will be angular components. (Which we honestly expected given part i). So when we take the derivative we'll get two componetns. First, the radial:

$$ E_{Ir} = \frac{Q}{4\pi\epsilon_0 r^2} + \frac{\epsilon Q R_2^2}{8 \pi \epsilon_0} \left[  \frac{2\epsilon_0}{3R_2^3} - \frac{2\epsilon}{3R_2^3} - \frac{1}{R_2^3} - \frac{2\epsilon_0}{3R_1^3} - \frac{\epsilon}{3R_1^3} \right] $$


However, $E_{I\theta}$ isn't so simple, for $\epsilon = \frac{\epsilon_0}{1-\alpha cos\theta}$. Fortunately the first term of the potnetial has no angles in it so it goes away, but $A_1$ does. 

$$ E_{I\theta} = r\frac{\partial}{\partial \theta}\frac{Q R_2^2}{8 \pi (1-\alpha cos\theta)} \left[  \frac{2}{3R_2^3} - \frac{2}{3R_2^3(1-\alpha cos\theta)} - \frac{1}{\epsilon_0R_2^3} - \frac{2}{3R_1^3} - \frac{1}{3R_1^3(1-\alpha cos\theta)} \right] $$

Ultimately this comes down to two derivatives, $\frac{1}{(1-\alpha cos\theta)}$ and $\frac{1}{(1-\alpha cos\theta)^2}$ The results...

$$ E_{I\theta} = r\frac{QR^2_2}{8\pi} \left[ -\frac{\alpha sin\theta}{(1-\alpha cos\theta)^2}\left( \frac{2}{3R_2^3} - \frac{1}{\epsilon_0 R_2^3} - \frac{2}{3R_1^3}\right)  + \frac{2\alpha sin\theta}{(1-\alpha cos\theta)^3}\left( -\frac{2}{3R^3_2} - \frac{1}{3R_1^3} \right) \right] $$


And naturally $E_{I\phi} = 0$ since we have no dependence there. 

Keep in mind this is all for region I, the other regions would have their own things (and require finding $A_2$). Do we think this answer is right? We highly suspect an algebra error somewhere considering just HOW MANY steps we took, but as for the method we think it is sound. Note that the angular dependence grows with r, this makes sense, after all the further away from the origin one gets the more relative distance each angle covers.
