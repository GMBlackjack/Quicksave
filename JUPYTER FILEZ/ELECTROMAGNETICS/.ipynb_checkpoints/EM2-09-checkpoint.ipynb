# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

So this is simlar to the free particle, except now that there are two forces. One force comes from the EM wave, and the other force comes from the spring. (We are assuming the produced field and magnetic field don't do anything, which comes from the "not large" amplitude of the incident wave. 

So, basically, we treat this as Thomson Scattering with an added force.

$$ \vec E_i = E_0 e^{i(kz-\omega t)} (\hat x + i\hat y)$$

Note the polarization is circular. We will be leaving off the i subscript for a while sicne it's the only E field we will be dealing with for a bit. 

Anyway, the forces on our paticle are $q\vec E$ and the one from the spring, $-k\vec r_0$ where we have chosen the vector $r_0$ to represent displacement. 

From this, we write the acceleration.

$$ \vec a = \frac{q}{m} \vec E - \vec{k}{m} \vec r_0 $$

Now, we can write this in the form of the current derivative.

$$ \frac{\partial j}{\partial t} = q\vec a = \frac{q^2}{m} \vec E - \frac{kq}{m} \vec r_0 $$



Next we want to take the integral of both sides with respect to time. However, we have a problem, and that problem is $\vec r_0$, the displacement. what form does it take in time? Well, we can actually assume an oscillation, as the displacmeent will be dictated by where the electric field is pulling it. However, we cannot just declare what the amplitude of this oscillation is, nor is it clear what direction it will be in--it will be related to the circular polarization, yes, but how exactly? 

Perhaps it will become obvious later. For now, we say $\vec r_0 = r_0 e^{i(kz-\omega t)} \hat r_0$

The exponential term must be related to the electric field's exponential, as that is ultimately the force that causes the displacement to exist in the first place. We shall leave our direction and amplitude as unknowns for now, perhaps later we shall be able to assign values to them.

Anyway, since we have noted the time dependence, we can now perform the integral.

$$ \vec j = \frac{q^2}{m} E e^{i(kz-\omega t)} \hat E - \frac{kq}{m} r_0 e^{i(kz-\omega t)} \hat r_0 dt  $$
$$ =\left[\frac{-1}{i\omega} \frac{q^2}{m} E e^{i(kz-\omega t)} \hat E +\frac{1}{i\omega} \frac{kq}{m} r_0 e^{i(kz-\omega t)} \hat r_0 \right] \delta(\vec r_1 - \vec r_0)  $$

The delta function reflects the fact that the point charge exists only at one point, its displacement, and is why we chose $\vec r_0$ to represent it. So we can now find the answer to this particular integral if we assume small perturbations at z=0. This does not mean that the particle is *at* the origin, merely that it is always *near* the origin. This is, yes, an approximation, but this doesn't remove much information. It just allows us to set the exponendtials to 1, and treat the delta function as existing *at* the origin. This also shouldn't destroy the cross section, as the motion should merely move the location of the cross seciton, not alter it's magnitude (at least not by much). 

$$ \int \vec j e^{i(\vec k \cdot \vec r)} d^3 r = -\frac{q^2}{i\omega m} E \hat E + \frac{kq}{i\omega m}r_0 \hat r_0  $$
$$ = \frac{q}{i\omega m} (k r_0 \hat r_0 - q E \hat E) $$

And we can now find the differential cross section.

$$ \frac{d\sigma}{d\Omega} = \frac{K^2}{(4\pi \epsilon_0 c E)^2} \left| \hat K \times \frac{q}{i\omega m} (k r_0 \hat r_0 - q E \hat E) \right|^2 $$

Where we are using K for the wavenumber since k is already taken for the spring constant. Anyway, the square absolute value gets entirely rid of the i and lets us pull out our leading constant. This cancels the K due to the presence of the frequency. We also pull out an extra factor of q and E, for reasons that will be obvious in a second. 

$$ \frac{d\sigma}{d\Omega} = \frac{q^2}{(4\pi \epsilon_0 m c^2)^2} \left| \hat K \times (\frac{k r_0}{qE} \hat r_0 - \hat E) \right|^2 $$

The leading coefficient is just the classical electron radius.

$$ \frac{d\sigma}{d\Omega} = r_e^2 \left| \hat K \times (\frac{k r_0}{qE} \hat r_0 - \hat E) \right|^2 $$ 

By the vector relation for square cross product magnitudes...

$$ = r_e^2 \left[ (\frac{k r_0}{qE} \hat r_0 - \hat E)^2 - (\hat K \cdot (\frac{k r_0}{qE} \hat r_0 - \hat E))^2 \right] $$ 

Now, to actually find the total cross section, we will need to integrate over all solid angles. The first term has no angular dependence, but the second term has it in relation to the K-unit-vector, which changes based on what angle we're looking at. We think we can take these integrals while leaving the actual directions entirely arbitrary, sicne we can take them over *all space*. So let's work things out and see what the resutls are...

$$ = r_e^2 \left[ \frac{k^2 r_0^2}{q^2E^2} \hat r_0^2 + \hat E^2 - \frac{2kr_0}{qE}(\hat r_0 \cdot \hat E) - \frac{k^2 r_0^2}{q^2E^2} (\hat K \cdot \hat r_0)^2 +2\frac{k r_0}{qE}( \hat K \cdot \hat E )(\hat K \cdot \hat r_0) - ( \hat K \cdot \hat E )^2 \right] $$

$$ = r_e^2 \left[ \frac{k^2 r_0^2}{q^2E^2} + 1 - \frac{2kr_0}{qE}(\hat r_0 \cdot \hat E) - \frac{k^2 r_0^2}{q^2E^2} +2\frac{k r_0}{qE}( \hat K \cdot \hat E )(\hat K \cdot \hat r_0) - 1 \right] $$

$$ = r_e^2 \left[- \frac{2kr_0}{qE}(\hat r_0 \cdot \hat E) +2\frac{k r_0}{qE}( \hat K \cdot \hat E )(\hat K \cdot \hat r_0) \right] $$

Now we must integrate over the solid angle, and now we care about the relation between $\hat r_0$ and $\hat E$. One thing we can say is that the force of E is driving the circular motion of the particle. Even though the spring force is also present, *it has to be acting in the same direction as E to produce circular motion*. For something to circulate there must be a total force that goes radially inward. The force of k and E point the same direction, however since this is an electron the actual orientation of E is opposite the direction of its force--but we must also remember that E is *opposing* the force of the spring, otherwise we would remain at the oriign not moving.. So E itself must point inward while its force points outward. Note that the radial vector poitns outward, thus the two unit vectors are paralell and pointing exactly opposite one another. This makes the dot product between the two of them simply -1, the standard result for a unit vector dotted with its own inverse.  

Perhaps more importantly, this makes $\hat r_0 = -\hat E$. Which we can use to simplify our relation even further.

$$ = r_e^2 \left[\frac{2kr_0}{qE} - 2\frac{k r_0}{qE}( \hat K \cdot \hat E )^2 \right] $$

Now, that second term is going to integrate to zero! $\hat K$ has to take every value over the solid integral, but $\hat E$ remains constant so every time we integrate over all direcitons we get terms that cancel out perfeclty. This leaves us with just the first term, which is simply a constant and thus gains a factor of 4$\pi$ when the integral over solid angle is performed.

$$ \sigma = r_e^2 \frac{8kr_0\pi}{qE} $$

Now, we have one last issue... the magnitude $r_0$, we have not yet determined what it is. But we *can*. The magnitude of the electric force acting on the particle in circular motion is $qE$, while the spring force is $kr_0$. $F=mr\omega^2$, and $\omega$ is provided from the frequency in E. Thus, 

$$ kr_0 - qE = mr_0\omega^2 $$

And we now need to solve for $r_0$

$$ r_0 = \frac{qE}{k - m\omega^2} $$

This makes our cross section...

$$ \sigma = r_e^2 \frac{8k\pi}{k - m\omega^2} $$

Notably independent of the electric field strength E, but we did assume small circulation so that makes sense. This is dependent on the frequency of light, though, which is to be expected. m in this case is the mass of the electron.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

So to find the dipole we start with the current. Since we have low frequency, at any given moment it's essentially like the sphere is experiencing a uniform electric field. 

$$ \vec j = \sigma \vec E $$. 

Where E is some wave with low frequency (and thus long amplitdue). From this we can find the dipole moment via the following relation.

$$ \int \vec j d^3r' = \frac{d\vec p}{dt} $$

At any given time the magnitude of E is constant throughout the sphere due to the long wavelength, so the integral is just the volume of the sphere times $\vec j$. 

$$ \frac43 \pi a^3 \sigma \vec E = \frac{d\vec p}{dt} $$

We have an incoming plane wave. To keep things simple, let's make it polarized in y and traveling in z.

$$ \frac43 \pi a^3 \sigma E \hat y e^{i(kz-\omega t)}= \frac{d\vec p}{dt} $$

And then we integrate to get the dipole.

$$ \frac43 \frac{i}{\omega} \pi a^3 \sigma E \hat y e^{i(kz-\omega t)}= \vec p $$

This notably changes over time. As it should, the oscillations will drag charges back and forth across the sphere. 

The equation for the absorption cross seciton is

$$ \sigma_{abs} = \frac{1}{<|S_{inc}|>} \int <\vec j \cdot \vec E> d^3r $$

This means that we're going to need the Poynting Vector of the incident wave. Since it's the time average, this is simply $\frac{E^2\hat k}{\mu_0 c}$, as it is for all EM plane waves. 

The current dotted into the electric field is effectively the electric field dotted into itself times the conductivity. So with just some basic adjustments, we already have...

$$ \sigma_{abs} = \frac{\mu_0 c}{E^2} \int \sigma < \vec E \cdot \vec E> d^3r $$

Note that taking the dot product *implies* taking the complex conjugate, so this is going to be real simple. 

$$ \Rightarrow \sigma_{abs} = \mu_0 c \sigma \int <1> d^3r $$

$$ \Rightarrow \sigma_{abs} = \mu_0 c \sigma \frac43 \pi a^3 $$

The optical theorem is

$$ \sigma_{tot} = \frac{4\pi}{k} \mathcal{I} \left[ \hat y \cdot \hat e  f(\theta = 0)\right] $$

(We have chosen y instead of x since that is our polarizaiton direction). 

We have the absorption cross section. We do not actually have the electric dipole scattering amplitude.However, as we are told Rayleigh Scattering, we are able to find it. First, take our electric dipole

$$ \frac43 \frac{i}{\omega} \pi a^3 \sigma E \hat y e^{i(kz-\omega t)}= \vec p $$

To find the scattering amplitude, we need to use this to find the $E_{scat}$, which is...

$$ E_{scat} = \frac{\mu_0}{4\pi r} \hat r \times (\hat r \times \vec p'') $$
$$ E_{scat} = -\frac{\mu_0}{4\pi} \omega^2 \frac43 \frac{i}{\omega} \pi a^3 \sigma E (\hat k \times (\hat k \times \hat y)) \frac{e^{i(kz-\omega t)}}{r} $$

$$ E_{scat} = -i \frac{\mu_0}{3\pi} \omega \pi a^3 \sigma E (\hat k \times (\hat k \times \hat y)) \frac{e^{i(kz-\omega t)}}{r} $$

Which means that the *scattering amplitude* is

$$ f = - i \frac{\mu_0}{3\pi} \omega \pi a^3 \sigma$$

And the scattering vector is 

$$ (\hat k \times (\hat k \times \hat y)) $$

Now this direciton can actually be simplified, since we want to deal wtih the direction at $\theta=0$, that is, the direction of propogation. In this case, that is z. 

$$ (\hat z \times (\hat z \times \hat y)) = - \hat y$$

Which if we plug into the optical theorem we get...

$$ \sigma_{tot} = \frac{4\pi}{k} \mathcal{I} \left[ \hat y \cdot \hat y (-1)(- i \frac{\mu_0}{3\pi} \omega \pi a^3 \sigma)\right] $$
$$  = \frac{4\pi}{k} \frac{\mu_0}{3\pi r} \omega \pi a^3 \sigma $$
$$  = \frac{4\mu_0}{3k} \omega \pi a^3 \sigma $$
$$  = \frac{4\mu_0}{3} c \pi a^3 \sigma $$

This is identical to what we got for the absorption cross section! Optical Theorem Satisfied!

It does seem somewhat odd, doesn't it, that when we used the Optical Theorem the "total" value of the cross seciton was just the "absorbed" cross section, huh? By simple algebra it would imply that since $\sigma_{tot} = \sigma_{abs} + \sigma_{scat}$ that $\sigma_{scat}$ would have to be zero! Except we clearly *have* a scattering wave, since we used the scattering amplitude to calculate the relation. What's going on here?

Well, why don't we actually try to examine what the scattering cross section is, at least qualitatively? We have a scattering electric field, let's try it out.

$$ E_{scat}^2 = \frac{\mu_0^2 \omega^2 a^3 \sigma^2 E^2}{9 r^2} |\hat k \times \hat y|^2  $$
$$ \frac{d\sigma_{scat}}{d\Omega} = \frac{\mu_0^2 \omega^2 a^3 \sigma^2}{9} |\hat k \times \hat y|^2  $$

Now we don't even have to take the integral to qualitatively examine this, the integral will be of order volume anyway since it's just the unit vectors that vary. No, what we note here is that the cross section scales with $\omega^2$. And $\omega$ was asssumed to be *small* in our approximations, so $\omega^2$ is *really* small. Thus, the scattering cross section exists at a higher order term than the approximation we found above. 

This indicates that our object is *mostly* absorbing, and scattering little. However, as the Optical Theorem makes plainly obvious, it is impossible to have a *perfectly* absorbing object. (Except in approximations that don't go out to enough terms. The Optical Theorem is not an approximation, but "small $\omega" is).
