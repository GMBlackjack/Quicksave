# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$



Fortunately for us we actually know what the answers have to be ahead of time. Inside a cavity, the electric field is always zero. Outside anything spherically symmetric, the electric field acts as a point charge. These are the results we should get. 

Regardless, Example 3.1 provides us the following for the electric field of a ring *along the z axis.*

$$ \textbf{E}(z\geq 0) = \frac{Q}{4\pi \epsilon_0} \frac{z}{(z^2+R^2)^{3/2}}\hat z $$

Now the initial plan, that of rotating the singular ring around to trace out the sphere, does not work. There are two reasons for this. One, we don't know the electric field anywhere but on the z axis. Two, by doing that no matter how we rotated the ring there would be a difference in density around the "poles" of rotation compared to the "equator."

So instead we model the sphere as a stack of rings of various sizes. What we essentially need to do is add up all the electric field contributions from each of the rings. Each ring has a variable R and a variable location. Introduce a new variable, z', that we integrate over while allowing z itself to remain. We will be integrating z' from -R to R (that is, the height of the sphere). z itself needs to be replaced with z-z' in order to capture the variability--naturally when z'=0, the original equation is recovered. R itself varies as well. An equation of a circle can be written as $z'^2+y^2=R^2$, so if we have z and want y we use $y = \sqrt{R^2-z'^2}$. Note we use the positive root since we want our current radius to be positive. 

Adjusting the charge density turned out not to be so simple, because the usual method of just replacing Q with $2 \pi \lambda R$ and then $2 \pi \sigma R$ does not work since *the size of the rings are changing*. Thus, their distance relative to each other is NOT directly related to z', which is genreally implicit in what we do. For dQ = $2 \pi \lambda dz'$ for the individual ring, but on the sphere it converts to $dQ = 2\pi \sigma dw$, where dw is the width of ring, which is not just z', for the sphere moves at an angle. In this case, $dw^2 = dz'^2 + dr^2$, which is to say the change in z' and the radius. Fortunately the change in radius is dependent on the change in z,since we already found $y = \sqrt{R^2-z'^2}$. Thus, $dr = \frac{-z'}{\sqrt{R^2-z'^2}} dz'$. This then means that $dw^2 = dz'^2 + \frac{z'^2}{R^2-z'^2} dz'^2 = \frac{R^2}{R^2-z'^2} dz'^2 $

Which means that our transformation beocmes $dQ = 2 \pi \sigma \frac{R}{\sqrt{R^2-z'^2}}dz' $

Let's do the case for outside the sphere first as that can be done in a single integral.

$$ \textbf{E}(z\geq 0) = \frac{\sigma}{2\epsilon_0} \int_{-R}^R \frac{(z-z')(\sqrt{R^2-z'^2})}{((z-z')^2+R^2-z'^2)^{3/2}}\frac{R}{\sqrt{R^2-z'^2}}\hat z dz' $$

To put it simply this is going to be a little awkward. However, we can simplify. 

$$ = \frac{\sigma}{2\epsilon_0} \int_{-R}^R \frac{(z-z')R}{(z^2 - 2zz' +R^2)^{3/2}}\hat z dz' $$

$$ = \frac{\sigma}{2\epsilon_0} \left[ \int_{-R}^R \frac{zR}{(z^2 - 2zz' +R^2)^{3/2}}\hat z dz' - \int_{-R}^R \frac{z'R}{(z^2 - 2zz' +R^2)^{3/2}}\hat z dz' \right] $$

Now these integrals can be evaluated. Take the one on the left first, plug it into Geogebra to get...

$$ \left[ \frac{R}{\sqrt{z^2-2zz'+R^2}} \right]_{-R}^{R} = \frac{R}{\sqrt{z^2-2zR+R^2}} - \frac{R}{\sqrt{z^2+2zR+R^2}}\\= \frac{R}{(z-R)} - \frac{R}{(z+R)} = \frac{2R^2}{(z-R)(z+R)} = \frac{2R^2}{z^2-R^2}$$

The second one is a bit uglier, using the integral-calculator from https://www.integral-calculator.com/

$$ \left[ \frac{R(-zz'+z^2+R^2)}{z^2\sqrt{z^2-2zz'+R^2}} \right]_{-R}^R $$
$$ = \frac{R(-zR+z^2+R^2)}{z^2\sqrt{z^2-2zR+R^2}} - \frac{R(zR+z^2+R^2)}{z^2\sqrt{z^2+2zR+R^2}}  $$
$$ = \frac{R(-zR+z^2+R^2)}{z^2(z-R)} - \frac{R(zR+z^2+R^2)}{z^2(z+R)}  $$
$$ = \frac{R((z-R)^2+zR)}{z^2(z-R)} - \frac{R((z+R)^2-zR)}{z^2(z+R)}  $$
$$ = \frac{R(z-R)}{z^2} - \frac{R(z+R)}{z^2} + \frac{R(zR)}{z^2(z-R)} - \frac{R(-zR)}{z^2(z+R)}  $$
$$ = \frac{-2R^2}{z^2} + \frac{R^2}{z(z-R)} + \frac{R^2}{z(z+R)}  $$
$$ = \frac{-2R^2}{z^2} + \frac{2R^2z}{z(z^2-R^2)}  $$
$$ = \frac{-2R^2}{z^2} + \frac{2R^2}{z^2-R^2}  $$

Now when the second integral is subtracted from the first, we get a neat result.

$$ \frac{\sigma}{2\epsilon_0} \frac{2R^2}{z^2} $$

Note that the total charge is $4\pi\sigma R^2$, which allows us to transform it into:

$$ \boxed{\frac{Q}{4\pi\epsilon_0} \frac{1}{z^2}} $$

Which is exactly what it should be outside.

Now immediately we think the inside will be a pain, but remember, it has to go to zero, so there's likely a funky trick. 

One way to deal with the inside is patially integrate two spherical shells and stop at some point. That is to say we would integrate from -R to z, then z to R with the sign flipped, since z is now inside that range somewhere. We can actually write this out rather easily. There is some awkardness in the bounds of the second part, but we can be sure that the result is just the negative version of what it would be otherwise, the only trick is that we are effectively MOVING z elsewhere, so we have to make it negative. 

$$ = \frac{\sigma}{2\epsilon_0} \left[ \int_{-R}^z \frac{zR}{(z^2 - 2zz' +R^2)^{3/2}}\hat z dz' - \int_{-R}^z \frac{z'R}{(z^2 - 2zz' +R^2)^{3/2}}\hat z dz' - \\ \int_{-R}^{-z} \frac{-zR}{(z^2 + 2zz' +R^2)^{3/2}}\hat z dz' + \int_{-R}^{-z} \frac{z'R}{(z^2 + 2zz' +R^2)^{3/2}}\hat z dz' \right] $$

Now, rather than evaluating each sum, we're going to examine the boundaries. What we expect is that the R-bounds will cancel and the z bounds will as well. We actually already have some idea about what the integrals turn into at the z'=R bounds, so let's write those down.

Term 1: $-\frac{R}{z+R}$

Term 2: $\frac{R((z+R)^2-zR}{z^2(z+R)}$

Term 3: $-\frac{R}{z-R}$

Team 4: $\frac{R((z-R)^2+zR)}{z^2(z-R)}$

Terms 1 and 3 combine to make $-\frac{2Rz}{z^2-R^2}$

Terms 1 and 4 combine to make $\frac{2R}{z} + \frac{2R^3}{z(z^2-R^2)}$

Now if we make them all of like terms, they DO cancel!

$$ -\frac{2Rz^2}{z(z^2-R^2)} + \frac{2R(z^2-R^2)}{z(z^2-R^2)} + \frac{2R^3}{z(z^2-R^2)} = 0$$

Which means we need to get all the z terms to cancel as well. z'=z or z'=-z, depending on which side we're evaluating. 

Term 1: $\frac{R}{\sqrt{R^2-z^2}}$

Term 2: $-\frac{R^3}{z^2\sqrt{R^2-z^2}}$

Term 3: $-\frac{R}{\sqrt{R^2-z^2}}$

Term 4: $\frac{R^3}{z^2\sqrt{R^2-z^2}}$

Term 1 cancels with term 3, term 2 cancels with term 4, meaning *everything* goes to zero, which means the inteior electric field is zero.

Which it should be. 

The disc is given to us by:

$$ \textbf{E}(z>0) = \frac{\sigma}{2\epsilon_0}\left[ 1 - \frac{z}{\sqrt{z^2+R^2}}\right] \hat z $$

Again, this is above the disc, below is equal but opposite. 

Unlike the previous problem which has the issue of integrating through another dimension, here we only want to find the field on the z-axis, which means we don't need any extra information as every "disc" that makes up the cylinder only has contributions along the central axis due to symmetry, we don't need to know the field anywhere else to know how it adds. 

What we need to do is integrate over the length, that is, to add up every contribution from each "Disc" at every point from 0 to L. To do this we need to first realize that sometimes we will have a position with discs above *and* below, so the z>0 limitation needs to be removed--when dealing with a point inside the cylinder. Outside the cylinder, that is z>L, we can work without doing this. However, what are we integrating over in the equation? Effectively, it's distance from z to whatever disc we're considering at that moment. Which means, effectivley, we need to add a new variable: call it l. z is the position of our particle. l is the number we vary to get the result we need. When l=0, the "z" needs to just be z. But when l=L, the "z" needs to be z-L. So we end up with...

$$ \textbf{E} (z>L) = \int_0^L \frac{\rho}{2\epsilon_0} \left[ 1 - \frac{z-l}{\sqrt{(z-l)^2+R^2}}\right] \hat z dl$$

This will be fun. First of all, this is going to be three integrals. 

$$ = \frac{\rho}{2\epsilon_0} \left[ \int_0^L  1 dl - \int_0^L \frac{z}{\sqrt{(z-l)^2+R^2}}dl + \int_0^L \frac{l}{\sqrt{(z-l)^2+R^2}}dl  \right] \hat z $$

$$ = \frac{\rho}{2\epsilon_0} \left[ L - z\int_0^L \frac{1}{\sqrt{(z-l)^2+R^2}}dl + \int_0^L \frac{l}{\sqrt{(z-l)^2+R^2}}dl  \right] \hat z $$

We turn to GeoGebra to solve the integrals since simplificaiton doesn't look like it will be simple. (Expanding the binomial ends up with a polynomial, not obviously simpler.) 

$$ = \frac{\rho}{2\epsilon_0} \left[ L - z\left[ -ln(z-l + \sqrt{(z-l)^2+R^2} \right]_0^L + \left[ -zln(z-l + \sqrt{(z-l)^2+R^2} + \sqrt{(z-l)^2+R^2} \right]_0^L  \right] \hat z $$

Note that, very conveniently, that the zln(stuff) expressions cancel with each other, meaning we only have to evaluate one thing.

$$ = \frac{\rho}{2\epsilon_0} \left[ L + \left[ \sqrt{(z-l)^2+R^2} \right]_0^L  \right] \hat z $$

$$ = \frac{\rho}{2\epsilon_0} \left[ L + \sqrt{(z-L)^2+R^2} -\sqrt{z^2+R^2} \right] \hat z $$

Which doesn't have an evident way to simplify, so we'll just leave it at that. 

$$ \boxed{\textbf{E}(z>L) = \frac{\rho}{2\epsilon_0} \left[ L + \sqrt{(z-L)^2+R^2} -\sqrt{z^2+R^2} \right] \hat z} $$

Now, the cool think is that this also works for z=L. Naturally, the contribution from one infinitesimal disc where the point *is* ceases to be, but a single infinitesimal disc can be removed with no influence on the overall state of the solid. (This can also be argued from a limit, we can get closer and closer to z=L from both ends this way.) So, If we jsut let Z=L, the contribution is:

$$ \textbf{E}(z=L) = \frac{\rho}{2\epsilon_0} \left[ L + R -\sqrt{L^2+R^2} \right] \hat z $$

This may not seem immediately helpful at first, after all, what we need is the electric field *inside* the cylinder. However, what is an electric field inside a cylinder but the combination of the electric field of two cylinders, one above and one below? the z < L case will just be two z=L cases added togehter with different values of L, specifically, z and L-z. Note that the L-z cylinder is resting on *top*, so its contribution is negative (the opposite directoin.)

$$ \textbf{E}(z<L) = \frac{\rho}{2\epsilon_0} \left[ (z + R -\sqrt{z^2+R^2}) - (L-z + R -\sqrt{(L-z)^2+R^2}) \right] \hat z $$
$$ \boxed{ \textbf{E}(z<L) = \frac{\rho}{2\epsilon_0} \left[ 2z-L -\sqrt{z^2+R^2} +\sqrt{(L-z)^2+R^2} \right] \hat z} $$

Which we will note is very similar to the equation for z>L, except with some terms flipped around. Which makes sense. When z=L/2 the interior solution vanishes, as it should. Also when z=0 we get the same solution as for z=L, except negative, also a good sign. Curiously, for really really large z, that is z >> L,R, the result for the exterior solution becomes $\frac{\rho L}{2 \epsilon_0}$, which indicates that we'd definitely need more refinement to get the $1/r^2$ behavior. However, the presence of the roots makes this not entirely obvious. 

Trying to look up the accepted value to check results in a lot of infinitely long cylinders concerned only with the radial portion of the field. Which is unhelpful.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$



First of all we note that, outside the charge distribution, we have a total charge of Q-Q = 0, so we're dealing with something electrically neutral. Furthermore, the limiting cases are going to be kind of extrene: an approaching-infinite charge density in the center, or an approaching-infinite charge density on the surface. That will be fun. 

Secondly, we were informed of a typo: the interior charge is -Q, the exterior os Q. This will do nothing but change the sign of the final answer, as charges are interchangeable. But we will follow it--interior negative, exterior positive. 

We freely refer to the two sections as "interior" and "exterior" since anything further out than $R_2$ must have no electric field at all as the system is electrically neutral. From lecture (and 3.80) we know the actuala total electrostatic energy is given by...

$$ U_E = \frac{\epsilon_0}{2} \int |\textbf{E}|^2 d^3r $$



Now in our case we can simplify this a bit. First of all, we're integrating over a spherically symmetric situation, so the volume integral will become a line integral with a factor of $4\pi$. Furthermore since we only have one direction, the electric field can be treated as a scalar, not a vector, with direction being encoded by the sign.

$$ U_E =2\pi\epsilon_0 \int |E(r)|^2 dr $$

Now, by Gauss' law, E(r) is equal to a point charge at any point based on the enclosed charge. 

$$ E(r) = \frac{1}{4\pi \epsilon_0}\frac{Q_{enc}}{r^2}$$

WHich makes our full equation...

$$ U_E = \frac{1}{8\pi \epsilon_0} \int (\frac{Q_{enc}}{r^2})^2 dr $$

Now, the enclosed charge depends on r, which is where most of the complecity of this problem arises. There are two distinct regions of behavior for Q: from 0 to $R_1$ and from $R_1$ to $R_2$. (Beyond this there is nothing). At those points, the charge is related to how much of the total volume of the section is currently enclosed. 

For the core this is relatively simple.

$$ -Q(\frac{r^3}{R_1^3})$$

Starts out at nothing but slowly approaches the full charge. Note the lack of $\frac 43 \pi$ that is usually present in spherical volumes--the thing is, we only care about a percentage here, and the percentage scales off the deifference in radius. There is no factor of $\pi$ since we are scaling off of the goal, that is, Q. Also it's negative since the interior charge is -Q. 

For the rest, we add the charge contribution from the exterior to a static -Q enclosed. 

$$  Q(\frac{r^3-R_1^3}{R_2^3-R_1^3}) -Q $$

Here the fraction is constructed in the same way, however we note the oddity of calculating the percent volume of the shell: the volume is always calculated by finding the volume of a sphere and then subtracting the interior sphere which, in this case, has a radius of $R_1$. So, now we can return to our total energy integral and evaluate it. It will certainly be fun.

$$ U_E = \frac{1}{8\pi \epsilon_0} \left[ \int_0^{R_1} (\frac{Q_{interior}}{r^2})^2 dr + \int_{R_1}^{R_2} (\frac{Q_{exterior}}{r^2})^2 dr \right] $$
$$ = \frac{1}{8\pi \epsilon_0} \left[ \int_0^{R_1} (\frac{-Q(\frac{r^3}{R_1^3})}{r^2})^2 dr + \int_{R_1}^{R_2} (\frac{Q(\frac{r^3-R_1^3}{R_2^3-R_1^3}) -Q}{r^2})^2 dr \right] $$
$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \int_0^{R_1} (-\frac{r}{R_1^3})^2 dr + \int_{R_1}^{R_2} ((\frac{r^3-R_1^3}{r^2(R_2^3-R_1^3)}) -\frac{1}{r^2})^2 dr \right] $$

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \int_0^{R_1} (-\frac{r}{R_1^3})^2 dr + \int_{R_1}^{R_2} (\frac{(r^3-R_1^3)^2}{r^4(R_2^3-R_1^3)^2}) -\frac{2(r^3-R^3_1)}{r^4(R^3_2-R^3_1)} +\frac{1}{r^4} dr \right] $$

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \int_0^{R_1} \frac{r^2}{R_1^6} dr + \int_{R_1}^{R_2} (\frac{(r^3-R_1^3)^2}{r^4(R_2^3-R_1^3)^2}) -\frac{2(r^3-R^3_1)}{r^4(R^3_2-R^3_1)} dr + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right) \right] $$

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \int_0^{R_1} \frac{r^2}{R_1^6} dr + \int_{R_1}^{R_2} (\frac{(r^3-R_1^3)^2}{r^4(R_2^3-R_1^3)^2}) dr - ln(\frac{R_2}{R_1})\frac{2}{(R^3_2-R^3_1)} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right)\frac{2R^3_1}{R^3_2-R^3_1} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right) \right] $$

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \frac{1}{3R_1^3} + \int_{R_1}^{R_2} (\frac{r^2}{(R_2^3-R_1^3)^2}) dr  - ln(\frac{R_2}{R_1})\frac{2R^3_1}{(R^3_2-R^3_1)^2} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right)\frac{R^6_1}{(R^3_2-R^3_1)^2}\\ - ln(\frac{R_2}{R_1})\frac{2}{(R^3_2-R^3_1)} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right)\frac{2R^3_1}{R^3_2-R^3_1} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right) \right] $$

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \frac{1}{3R_1^3} + \frac{1}{3(R_2^3-R_1^3)}  - ln(\frac{R_2}{R_1})\frac{2R^3_1}{(R^3_2-R^3_1)^2} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right)\frac{R^6_1}{(R^3_2-R^3_1)^2}\\ - ln(\frac{R_2}{R_1})\frac{2}{(R^3_2-R^3_1)} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right)\frac{2R^3_1}{R^3_2-R^3_1} + \frac 13\left( \frac{1}{R_1^3} - \frac{1}{R_2^3} \right) \right] $$

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \frac{1}{3R_1^3} + \frac{1}{3(R_2^3 - R_1^3) } - ln(\frac{R_2}{R_1})\frac{2R^3_1}{(R^3_2-R^3_1)^2} - ln(\frac{R_2}{R_1})\frac{2}{(R^3_2-R^3_1)} + \frac13 \left( \frac{1}{R^3_1} - \frac{1}{R_2^3} \right)\left( \frac{R^6_1}{(R_2^3-R^3_1)^2}+ \frac{2R_1^3}{R^3_2-R^3_1} +1 \right) \right] $$

Which is... technically an answer, but one thinks that surely there must be a way to simplify it. We don't really see a way to do that, though. We also went through and re-did all the integrals manually, nothing has changed. 

Now, even if this isn't exactly simplified, we *CAN* still examine the extreme behaviors. For instance, when $R_1 \rightarrow 0$, that is, when the core approaches becoming an infinitesimal charge, the total energy simplifies to:

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \frac{1}{3R_1^3} + \frac{1}{3(R_2^3) } - ln(\frac{R_2}{R_1})\frac{2}{(R^3_2)} + \frac13 \left( \frac{1}{R^3_1}\left( \frac{2R_1^3}{R^3_2}\right) - \frac{1}{R_2^3} \right) \right] $$
$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \frac{1}{3R_1^3} - ln(\frac{R_2}{R_1})\frac{2}{(R^3_2)} + \frac13 \frac{1}{R^3_1}\left( \frac{2R_1^3}{R^3_2}\right)\right] $$

Limits are always annoying to work with, case in poitn being the second term here which is of the form $R^3/R^3$, an indeterminate form. In any situation where the powers of growth match, though, we end up with the constant that remains. 

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[ \frac{1}{3R_1^3} - ln(\frac{R_2}{R_1})\frac{2}{(R^3_2)} + \frac{2}{3R^3_2}\right] $$

We did in fact check this limit with GeoGebra to make sure we weren't remembering how this worked wrong, or missed a nuance in the previous simplificaiton steps. Regardless, the natural log term clearly diverges to positive infinity, matching what we would expect for an infinite charge density at the center of a sphere--explosive increase of potential in the system. The leading fraciton also diverges. 

Now consider the case when the two radii are equal. This makes a few obvious simplifications, and all natural log terms become 0.

$$ = \frac{Q^2}{8\pi \epsilon_0} \left[\frac{1}{3R_1^3} + \frac{1}{3(R_2^3 - R_1^3) } \right] $$

This, too, diverges to infinity. But we were expecting it not to... possible calculation error, since in-class it was implied that there shoudl be a value here that doesn't diverge. After all, charges congregate on the edge of a conductor all the time.
