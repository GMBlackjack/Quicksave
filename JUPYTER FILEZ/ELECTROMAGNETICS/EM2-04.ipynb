# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Well we can tell right away that at z=0, the wave is linearly polarized along the x-axis, since that's where we start. THe question now is what other locations does this occur?

To figure this out we need to split up the x polarization into a sum of left circular and right circular waves. Since we don't exactly care which is which we only care that we can tell them apart, we shall simply call associate the two circular polarizaitons with "negative imaginary" and "positive imaginary", that is...

$$ \hat x = \frac 12(\hat x + i \hat y) + \frac12(\hat x - i\hat y) $$

At z=0 these two polarizations add directly to produce the pure x polarization.

$$ \vec E(0,t) = \frac12 E_0 \left[ (\hat x + i\hat y) e^{i\omega t} + (\hat x - i\hat y) e^{i\omega t} \right] $$

There is a supressed kz term, let's add that back in. 

$$ \vec E(0,t) = \frac12 E_0 \left[ (\hat x + i\hat y) e^{i(kz + \omega t)} + (\hat x - i\hat y) e^{i(kz+\omega t)} \right] $$

Note that its $+\omega t$ rather than $-\omega t$. We know this since the imaringary exponent is positive rather than negative. Regardless, to actually add the z dependence back, we need to insert the indeces of refraction. 

$$ \vec E(z,t) = \frac12 E_0 \left[ (\hat x + i\hat y) e^{i(n_Lz\omega/c + \omega t)} + (\hat x - i\hat y) e^{i(n_Rz\omega/c+\omega t)} \right] $$


Our goal is to find every z value where the polarization is purely x, which means the y directional components need to cancel entirely. This will obviously occur when the two exponentials are equal to each other--for all *time*. The fact that time has to be consistent is a little annoying here. However, to simplify that, we can just pull OUT the time dependence since these are exponentials.

$$ \vec E(z,t) = \frac12 E_0 e^{i\omega t}\left[ (\hat x + i\hat y) e^{in_Lz\omega/c} + (\hat x - i\hat y) e^{in_Rz\omega/c} \right] $$

And it makes more sense that these z-depenmdent portions will occasionally cycle to be equal to each other. Let's see if we can solve for z knowing this.

$$e^{in_Lz\omega/c} = e^{in_Rz\omega/c} $$
$$\Rightarrow n_Lz\omega/c = n_Rz\omega/c + 2\pi m $$
Where m is some integer. If m=0, we get the z=0 situation. However, if we allow m to be any integer...
$$ \Rightarrow n_L\omega/c = n_R\omega/c + 2\pi m / z  $$
$$ \Rightarrow \frac{\omega(n_L+n_R)/c}{2\pi m} = 1 / z  $$
$$ \Rightarrow \frac{2\pi m}{\omega(n_L+n_R)/c} = z  $$
So to find any position with x polarization we simply plug in an integer m starting from 0 and going to 1, 2, 3... etc. If the medium continues backward we can also use negative numbers. 

Naturally this is only true if $n_L \neq n_R$, because then the two exponentials are always the same and the light remains polarized in x everywhere.


-45 degree polarization is given by $\hat x - \hat y$. In order to create this, we would need to multiply the two circular polarizations by some complex number that arrives at that result. We suspect that number does not exist, but let's prove it.

$$ (\hat x + i\hat y)(a_1+ib_1) + (\hat x - i\hat y)(a_2+ib_2) $$
$$ =  \hat x (a_1+ib_1 + a_2+ib_2) + (i\hat y)(a_1+ib_1 - a_2 - ib_2) $$

The x hat direction cannot be imaginary, so we need $b_1 = -b_2$. If this is true, then on the y direction's side we end up with 2ib. Since we need the y direction's real component to entirely cancel, that means $a_1 = a_2$. From this we update...

$$ =  \hat x (2a_1) + (i\hat y)(2ib_1) $$
$$ =  2 \hat x a_1 - 2 \hat y b_1 $$

Curious, we have found a combination that gives a -45 degree polarization if the magnitudes of a and b are the same. This means that what we need the exponentials to become is...

$$ (\hat x + i\hat y)(a+ia) + (\hat x - i\hat y)(a-ia) $$

Now, can this actually be accomplished? Can we find a z that makes both of these true at once? Well, remember that we can split up our exponentials into trig funcitons...

$$ (\hat x + i\hat y)(cos(n_Lz\omega/c) + isin(n_Lz\omega/c)) + (\hat x - i\hat y)(cos(n_Rz\omega/c) + isin(n_Rz\omega/c)) $$

To simplify, consider the interiors of the trig functions a single variable, let it be X and Y.

$$ (\hat x + i\hat y)(cos(X) + isin(X)) + (\hat x - i\hat y)(cos(Y) + isin(Y)) $$

We need $sin(Y) = -sin(X)$ and $cos(X) = cos(Y)$, which are two equaitons and two unknowns. There are actually several of these points, say, $X=-\pi/2, Y=\pi/2$. Sign change does nothing to a cosine, but it flips a sine. This poitn corresponds to

$$ -\pi /2 = n_Lz\omega/c \Rightarrow z = -\pi / 2n_L \omega/c $$
$$ \pi /2 = n_Rz\omega/c \Rightarrow z = \pi / 2n_R \omega/c $$

This seems to indicate that z has to be in two places at once for the polarization we seek to be true, which is physically impossible without some conditions on the indeces of refraction themselves. (Such as allowing one to go negative). 

To remove all shadow of a doubt let's look at the *entire* solution set of the system of equations, then. 

$$ X = m\pi/2 $$
$$ Y = -X + c2\pi $$

Where c is another integer since we're already using n and i and j have other meanings. 

With this, we get...

$$ z = m\pi/2n_L \omega/c $$
$$ z = -m\pi/2n_R \omega/c + c2\pi/n_R \omega/c  $$

Which implies

$$ z = -z \frac{n_L}{n_R} + \frac{c2\pi}{n_R \omega/c}  $$
$$ \Rightarrow 1 + \frac{n_L}{n_R} = \frac{c2\pi}{z n_R \omega/c}  $$
$$ \Rightarrow \frac{n_R \omega/c}{c2\pi}(1 + \frac{n_L}{n_R}) = \frac{1}{z}  $$
$$ \Rightarrow \frac{c2\pi}{n_R \omega/c}(1 + \frac{n_L}{n_R})^{-1} = z  $$

Which means that yes, there *do* exist locaitons where we can get this polarization! If we let c=1...

$$ \Rightarrow \frac{2\pi}{n_R \omega/c}(1 + \frac{n_L}{n_R})^{-1} = z  $$

Of course, we should check this since our intuition was that such a point was not possible. We checked it by plugging it back into the trig functions...

$$ sin(n_R \omega/c \frac{2\pi}{n_R \omega/c}(1 + \frac{n_L}{n_R})^{-1}) = -sin(n_L \omega/c\frac{2\pi}{n_R \omega/c}(1 + \frac{n_L}{n_R})^{-1}) $$
$$ \Rightarrow sin(2\pi(1 + \frac{n_L}{n_R})^{-1}) = -sin(2\pi \frac{n_L}{n_R}(1 + \frac{n_L}{n_R})^{-1}) $$

And since this is just a check and not a proof, we just plugged some values into a computer to see. No matter what the n values are, this equality still holds. (Even if n goes negative!)

The relation also works for the cosine version.

So to our surprise, the -45 degree polarizaiton does, in fact, exist. And in the process of proving it existed we found all of them, just choose an integer c. 




This one is a little easier to understand. Since each of the polraizations is keyed to a ciruclar polarizaiton, we just need one of them to be zero while the other is purely real.

$$ (\hat x + i\hat y) e^{in_Lz\omega/c} + (\hat x - i\hat y) e^{in_Rz\omega/c} $$

In this case the $n_R$ term needs to be zero. But wait... that can't be done with an imaginary exponential! The cosine and sine components are never zero at the same time, so the exponential either provides a real or imaginary part, which keeps the polarization from ever being pure in any circular direction!

It seems as if our intuition was backward, for we expected b) not to exist and c) to be rahter simple to find, and yet b) turns out to actually be real and c) utterly impossible due to the nature of imaginary exponentials. Interesting!

So since no angles are even mentioned, will assume that the waves are going directly for the interface, that is, the angle of incidence is 0, and thus the angles of reflection and transmission are also 0. 

We are told they have linear polarization but not the *same* linear polarization, naturally they will have to be the same polarization to cancel in any meaningful way, but we will only specify this at the end.

One way we can think of to handle this is to take each incoming electric field separately, find its transmitted and reflected waves that result from it, and then add everything together. The simplest way to do this is to find the indeces of reafraction. 

$n = \sqrt{\frac{\mu\epsilon}{\mu_0\epsilon_0}}$. In our case $\mu=\mu_0$ so this reduces to $n = \sqrt{\frac{\epsilon}{\epsilon_0}}$. Therefore...

$$ n_1 = \sqrt{\frac{\epsilon_1}{\epsilon_0}}; n_2 = \sqrt{\frac{\epsilon_1}{\epsilon_0}}$$

Linearly polarized light heading dead-on to a flat boundary is neither strictly TE or TM, it depends on which direction we look at the polarization from. (That said in this case the only difference is a sign). We choose TE with E pointing out of the page. 

This means that we have our various relations true for each case:

$$ \frac{E_R}{E_I} = \frac{Z_2 cos\theta - Z_1 cos\theta_2}{Z_1cos\theta_1 + Z_2 cos\theta_2} = \frac{Z_2 - Z_1}{Z_1 + Z_2}$$
$$ \frac{E_T}{E_I} = \frac{2Z_2 cos\theta_2}{Z_1cos\theta_1 + Z_2 cos\theta_2} = \frac{2Z_2}{Z_1 + Z_2}$$

If we used TM the sign of the reflective ratio would change. 

Anyway, we now need the impedence. This is just $\sqrt{\mu/\epsilon} = \sqrt{\mu_0/\epsilon} $ for each material, so to save space we shall leave it as Z. (This is only true for dielectrics, its more complicated for conductors, but that's not our concern at the moment). 

Anyway, we can now find the relative magnitudes of each individual wave's reflections and transmissions. 

$$ E_{R1} = E_1 \frac{Z_2 - Z_1}{Z_1 + Z_2}  $$
$$ E_{T1} = E_1 \frac{2Z_2}{Z_1 + Z_2}  $$
$$ E_{R2} = E_2 \frac{Z_1 - Z_2}{Z_1 + Z_2}  $$
$$ E_{T2} = E_2 \frac{2Z_1}{Z_1 + Z_2}  $$

Note that the index on the Z terms flipped since the second wave is going from medium 2 to medium 1. 

Anyway, now every side of the situation has *three* waves. Assuming they all have the same polarization (the only way to get them to simply add) the outgoing magnitudes are...

$$ E_1 \frac{Z_2 - Z_1}{Z_1 + Z_2} + E_2 \frac{2Z_1}{Z_1 + Z_2} $$ 
$$ E_1 \frac{2Z_2}{Z_1 + Z_2} + E_2 \frac{Z_1 - Z_2}{Z_1 + Z_2} $$

We want no outgoing wave, so we take the parts of the wave in section 2 and set them to zero.

$$ E_1 \frac{2Z_2}{Z_1 + Z_2} + E_2 \frac{Z_1 - Z_2}{Z_1 + Z_2} = 0$$
$$ \Rightarrow E_1 \frac{2Z_2}{Z_1 + Z_2} = - E_2 \frac{Z_1 - Z_2}{Z_1 + Z_2} $$
$$ \Rightarrow E_1 2 Z_2 = - E_2(Z_1 - Z_2)$$
$$ \Rightarrow - \frac{2 Z_2}{Z_1 - Z_2} = \frac{E_2}{E_1}$$
$$ \Rightarrow \frac{2 Z_2}{Z_2 - Z_1} = \frac{E_2}{E_1}$$

Which is what the problem sought. However, we are curious what this situation means for medium 1. (For medium 2 it just means $-E_2$ is all that's there). 

$$ E_2 = E_1 \frac{2 Z_2}{Z_2 - Z_1} $$

So in medium 1 the outgoing wave is...

$$ E_1 \frac{Z_2 - Z_1}{Z_1 + Z_2} +  E_1 \frac{2Z_2}{Z_2 - Z_1} \frac{2Z_1}{Z_1 + Z_2} $$ 
$$ = E_1 \frac{Z_2 - Z_1}{Z_1 + Z_2} +  E_1 \frac{4Z_1Z_2}{Z_1^2 + Z_2^2} $$ 

Proving that there is still an outgoing wave in medium 1, despite there being none in medium 2.

We shall prove that this is true in general. 

First of all, it's trivially true if the polarizations don't match up, the waves can be considered entirely different components and we already know that power matches for individual waves hitting an interface. The problem presumably wants us to show it for the entire system at once. Rather than do it in just the medium 2 outgoing = 0 case, we shall do it in general (assuming same polarizaiton).

Incoming waves
$$ E_1 $$
$$ E_2 $$

Outgoing waves
$$ E_1 \frac{Z_2 - Z_1}{Z_1 + Z_2} + E_2 \frac{2Z_1}{Z_1 + Z_2} $$ 
$$ E_1 \frac{2Z_2}{Z_1 + Z_2} + E_2 \frac{Z_1 - Z_2}{Z_1 + Z_2}   $$

These are the magnitudes of our fields. In general, we wish to show the conservation of power, so each side needs to evaluate to the same thing. Power is given by the Poynting vector, which in this case simplifies rapidly to $E^2/Z$.

There are a few ways to run this calculation, we can try to make the ratio equal 1 or get all the terms to cancel to 0, which in the end prove the same thing but the methodology for each is a bit different. However, we can make a helpful observation: electric fields are always additive, as are magnetic fields. The power carried by one wave is independent of the power carried by the others. Thus, we can handle wave 1 and 2 entirely separately, proving that the power in and power out is the same for each. 

One could argue that since power is based on the Poynting vector, we cannot simply add things--but that's only the case for the components of each individual wave, as they are coupled. Otherwise how could we compare the power from the incident wave to a reflective wave, since the two would fundamentally be related to each other! Thus, each wave provides its own power. (It also helps that we are currently only concerned with magnitudes). 

This also means that $E_1$ and $E_2$ should balacne separately. So we just use the ratios for single waves...

$$ \frac{E_R^2}{E_I^2} \frac{Z_1 cos\theta_1}{Z_1 cos\theta_1} + \frac{E_T^2}{E_I^2} \frac{Z_1 cos\theta_2}{Z_2 cos\theta_1} = 1 $$

With our angles of zero this zimply becomes

$$ \frac{E_R^2}{E_I^2} + \frac{E_T^2}{E_I^2} \frac{Z_1}{Z_2} = 1 $$

So if we use the values for $E_1$ we get...

$$ \frac{E_1^2 \frac{(Z_2 - Z_1)^2}{(Z_1 + Z_2)^2}}{E_1^2} + \frac{E_1^2 \frac{4Z_2^2}{(Z_1 + Z_2)^2}}{E_1^2} \frac{Z_1}{Z_2} = 1 $$
$$ \Rightarrow \frac{(Z_2 - Z_1)^2}{(Z_1 + Z_2)^2} + \frac{4Z_2^2}{(Z_1 + Z_2)^2} \frac{Z_1}{Z_2} = 1 $$
$$ \Rightarrow \frac{Z_2^2+Z_1^2-2Z_1Z_2}{(Z_1 + Z_2)^2} + \frac{4Z_2Z_1}{(Z_1 + Z_2)^2} = 1 $$
$$ \Rightarrow \frac{Z_2^2+Z_1^2-2Z_1Z_2}{(Z_1 + Z_2)^2} + \frac{4Z_2Z_1}{(Z_1 + Z_2)^2} = 1 $$
$$ \Rightarrow \frac{Z_2^2+Z_1^2+2Z_1Z_2}{(Z_1 + Z_2)^2} = 1 $$
$$ \Rightarrow \frac{(Z_1 + Z_2)^2}{(Z_1 + Z_2)^2} = 1 $$

Which is clearly true. 

The process for $E_2$ is virtually identical. 

$$ \frac{E_2^2 \frac{(Z_1 - Z_2)^2}{(Z_1 + Z_2)^2}}{E_2^2} + \frac{E_2^2 \frac{4Z_1^2}{(Z_1 + Z_2)^2}}{E_2^2} \frac{Z_2}{Z_1} = 1 $$
$$ \Rightarrow \frac{(Z_1 - Z_2)^2}{(Z_1 + Z_2)^2} + \frac{4Z_1^2}{(Z_1 + Z_2)^2} \frac{Z_2}{Z_1} = 1 $$
$$ \Rightarrow \frac{Z_2^2+Z_1^2-2Z_1Z_2}{(Z_1 + Z_2)^2} + \frac{4Z_2Z_1}{(Z_1 + Z_2)^2} = 1 $$
$$ \Rightarrow \frac{Z_2^2+Z_1^2-2Z_1Z_2}{(Z_1 + Z_2)^2} + \frac{4Z_2Z_1}{(Z_1 + Z_2)^2} = 1 $$
$$ \Rightarrow \frac{Z_2^2+Z_1^2+2Z_1Z_2}{(Z_1 + Z_2)^2} = 1 $$
$$ \Rightarrow \frac{(Z_1 + Z_2)^2}{(Z_1 + Z_2)^2} = 1 $$

So in both cases power is conserved, no matter what $E_2$ and $E_1$ are. Since power is conserved individually it must be conserved on the whole. 

So our full wave arrangement for part a) is...

Incoming waves
$$ E_1 $$
$$ E_2 $$

Outgoing waves
$$ E_1 \frac{Z_2 - Z_1}{Z_1 + Z_2} + E_2 \frac{2Z_1}{Z_1 + Z_2} $$ 
$$ E_1 \frac{2Z_2}{Z_1 + Z_2} + E_2 \frac{Z_1 - Z_2}{Z_1 + Z_2}   $$

180 degrees out of phase will swap the signs of all $E_2$ componetns since $cos(x-180^{\circ}) = -cos(x)$ So we now have:

Incoming waves
$$ E_1 $$
$$ -E_2 $$

Outgoing waves
$$ E_1 \frac{Z_2 - Z_1}{Z_1 + Z_2} - E_2 \frac{2Z_1}{Z_1 + Z_2} $$ 
$$ E_1 \frac{2Z_2}{Z_1 + Z_2} - E_2 \frac{Z_1 - Z_2}{Z_1 + Z_2}   $$

The ratio we found in part a) was

$$ \frac{2 Z_2}{Z_2 - Z_1} = \frac{E_2}{E_1}$$

Or...

$$ E_2 = E_1 \frac{2 Z_2}{Z_2 - Z_1} $$

We care about the outgoing wave in region 2, so let's make the substitution in region 2's outgoing waves.

$$ E_1 \frac{2Z_2}{Z_1 + Z_2} - E_1 \frac{2 Z_2}{Z_2 - Z_1} \frac{Z_1 - Z_2}{Z_1 + Z_2} $$
$$ \Rightarrow E_1 \frac{2Z_2}{Z_1 + Z_2} + E_1 \frac{2 Z_2}{Z_1 + Z_2} $$
$$ \Rightarrow E_1 \frac{4Z_2}{Z_1 + Z_2}  $$

And this is our answer. Rahter than having the terms cancel, they interfere constructively. Which makes sense, being 180 degrees out of phase tends to have effects like this, either complete deconstruciton or complete construction.
