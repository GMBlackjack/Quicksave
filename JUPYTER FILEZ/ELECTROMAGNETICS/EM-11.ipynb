# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Basically, what we do here is take the magnetic field along the axis of a wire loop and then integrate it over the entire length, -L/2 to L/2. 

$$ \vec B = \frac12 \mu_0 I \frac{R^2}{(z^2+R^2)^{3/2}}\hat z $$

This is the field for one loop, we just need to adjust it for being at any position. Let z be our position, and z' be the new variable.

$$ \vec B = \frac{\mu_0 N I}{2L}\int_{-L/2}^{L/2} \frac{R^2}{((z'-z)^2+R^2)^{3/2}} dz'\hat z $$
$$ = \frac{\mu_0 N I}{2L}\left[ \frac{z'-z}{\sqrt{(z'-z)^2+R^2}} \right]_{-L/2}^{L/2}\hat z $$
$$ = \frac{\mu_0 N I}{L}\left[ \frac{L/2-z}{2\sqrt{(L/2-z)^2+R^2}} - \frac{-L/2-z}{2\sqrt{(-L/2-z)^2+R^2}} \right] \hat z$$

Which, while a bit ugly, is of the correct form, as it is jsut z L R and is unitless. 

Now we calculate at z=0 and z=L/2 and z=-L/2.

$$ \vec B(0) = \frac{\mu_0 N I}{2\sqrt{(L/2)^2+R^2}} \hat z$$
$$ \vec B(L/2) = \frac{\mu_0 N I}{2\sqrt{L^2+R^2}} \hat z$$
$$ \vec B(-L/2) = \frac{\mu_0 N I}{2\sqrt{L^2+R^2}} \hat z$$

And so the ratio requested is 

$$ \frac{2\sqrt{(L/2)^2+R^2}}{2\sqrt{L^2+R^2}} $$

And in the large-L limit this is...

$$ \frac{1}{2} $$

Which is to say the field strength at the ends is precisely half of what it is in the center. Which isn't all that surprising, though it is convenient. 

In general, $\vec M = (\mu/\chi) \vec B$. Since we are assuming a pure z-direction magnetization, we are making the approximation that the axis magnetic field is constant throughout the rod. (Or that we have an extremely thin rod, that also produces the same result. Sensible since L>>R). 

We are also pretending that volume magnetic "charge" is nonexistent. which is an approximation that makes less sense, especially for a very large rod where the stuff in the middle will basically only encounter volume "charge," but let's roll with it. Anyway, with this, our magnetization is

$$ \vec M = \frac{\mu}{\chi_m} \frac{\mu_0 N I}{L}\left[ \frac{L/2-z}{2\sqrt{(L/2-z)^2+R^2}} - \frac{-L/2-z}{2\sqrt{(-L/2-z)^2+R^2}} \right] \hat z $$

Note that we are not provided $\mu$, but we can find it from $\chi_m$.

$$ \mu = \mu_0(1+\chi_m) $$

We will continue to write $\mu$ in the euqation, however, to keep things neat. 

From here we find the surface charge densities, which is just the magnetization dotted with the normal vector. 

$$ \pm \sigma = \pm \frac{\mu}{\chi_m} \frac{\mu_0 N I}{L}\left[ \frac{L/2-z}{2\sqrt{(L/2-z)^2+R^2}} - \frac{-L/2-z}{2\sqrt{(-L/2-z)^2+R^2}} \right] \hat z $$

Where the minus sign is associated with the -L/2 area and plus with L/2. Since these surfaces have fixed z, we can just replace the big mess with the values at L/2 and -L/2. Which are the same as we found out in part a. 

$$ \pm \sigma = \pm \frac{\mu}{\chi_m} \frac{\mu_0 N I}{2\sqrt{L^2 + R^2}} \hat z $$

Now the H-field can be attained next, which is just the altered electric field of two oppositely charged discs. Which becomes

$$ \vec H = \pm \frac{M(\pm L/2)}{2} \left[ 1 - \frac{|z|}{\sqrt{z^2+R^2}} \right] \hat z $$
$$ = \pm \frac{\mu}{\chi_m} \frac{\mu_0 N I}{4\sqrt{L^2 + R^2}} \left[ 1 - \frac{|z|}{\sqrt{z^2+R^2}} \right] \hat z $$

For one disc. We have two discs at different locaitons, so we have to be careful when adding them together. Not only are they opposing sign, but they are on opposing sides as well. We are *below* the top surfac eand *above* the bottom surface, offset by factors of L/2. This creates the rather interesting result of...

$$ \vec H = \frac{\mu}{\chi_m} \frac{\mu_0 N I}{2\sqrt{L^2 + R^2}} \left[ \frac{|L/2-z|}{2\sqrt{(L/2 - z)^2+R^2} } + \frac{|z + L/2|}{2\sqrt{(z+L/2)^2+R^2}} - 1\right] \hat z $$

The negative sign is out front since positive charges repel, thus everything goes *down*. Now this is VERY similar to the original function we had in M, so let's rework it until it IS the same form. First, we have z on a very specific range, so we can get rid of those absolute value signs. 

$$ \Rightarrow \vec H = \frac{\mu}{\chi_m} \frac{\mu_0 N I}{2\sqrt{L^2 + R^2}} \left[ \frac{L/2 - z}{2\sqrt{(L/2 - z)^2+R^2} } + \frac{z + L/2}{2\sqrt{(z+L/2)^2+R^2}} - 1\right] \hat z $$

And yes a positive and a negatively charged plate would add constructively. This does, in fact, match f(z) exactly, for the extra minus signs on the second term can easily be converted into pluses. 

$$ \Rightarrow \vec H = \frac{\mu}{\chi_m} \frac{\mu_0 N I}{2\sqrt{L^2 + R^2}} (f(z)-1) \hat z $$

Now one might thing we just use one of the other relations for linear paramagnetic materials:

$$ \vec B = \mu_0(1+\chi_m)\vec H $$

But keep in mind we have actually only found H for the bound charged up there, we need to add it to the original H-field as well (which is just $B_0/\mu_0$ since there's no M in the original system)

$$ \vec B = \mu_0(1+\chi_m)(\vec H_f + \vec H_b) $$

Now we're going to do something a little odd and consider our leading coefficient in the bound H-field as an unknown to be solved for. Since it matches exactly to what we want, we call it $\sigma^*$

$$ \vec B = (1+\chi_m)(B_0 + \mu_0 \sigma^* (f(z)-1)) \hat z $$

Now. What exactly is $\sigma^*$? We mean, we have it, it's

$$ \sigma^* = \frac{\mu}{\chi_m} \frac{\mu_0 N I}{2\sqrt{L^2 + R^2}} \approx \frac{\mu}{\chi_m} \frac{\mu_0 N I}{2L} $$

But that is decidedly not the form we want and algebra is a pain, so we're going to find it another way. The trick here is to realize that, yes, this is in fact the magnetic charge density that we found. We can thus use the other relation to create some gymnastics:

$$ \vec M = \chi_m \vec H \Rightarrow \sigma^* = \chi_m H(L/2) $$

It doesn't matter if we are on the positive or negative side, the relation will hold. We can relate this to...

$$  \sigma^* / \chi_m= B_0(L/2) / \mu_0 +  \sigma^* (f(L/2)-1) $$ 

No we don't particuarly think this step is an obviously sensible one, but it will come around. We can repalce values of $B_0$ and $f(z)$ with their evaluations at L>>R, which we have mostly done in part a).

$$ \Rightarrow \sigma^* / \chi_m= \frac{NI}{2L} + \sigma^* (1/2-1) $$ 

With some rearranging...

$$ \Rightarrow \sigma^* (\frac{1}{ \chi_m} + 1/2) = \frac{NI}{2L} $$ 
$$ \Rightarrow \sigma^* = \frac{\frac{NI}{2L}}{(\frac{1}{ \chi_m} + 1/2)} $$ 
$$ \Rightarrow \sigma^* = \frac{NI}{2L}\frac{1}{(\frac{1}{ \chi_m} + 1/2)} $$ 
$$ \Rightarrow \sigma^* = \frac{NI}{L}\frac{1}{(\frac{2}{\chi_m} + 1)} $$ 
$$ \Rightarrow \sigma^* = \frac{NI}{L}\frac{1}{(\frac{2+\chi_m}{\chi_m} )} $$ 
$$ \Rightarrow \sigma^* = \frac{NI}{L}\frac{\chi_m}{2+\chi_m} $$ 

Which is exactly what we sought. 

Not sure why we need it in this form and not in the original form, but all right. 

Theoretically this means we can say:

$$ \Rightarrow \sigma^* = \frac{NI}{L}\frac{\chi_m}{2+\chi_m} = \frac{\mu}{\chi_m} \frac{\mu_0 N I}{2L} $$ 

Though we are admittedly unsure how we would go about actually showing this. 

$$ \vec B(z) \approx (1+\chi_m)\left[ B_0(z) + \mu_0 \frac{NI}{L} \frac{\chi_m}{\chi_m+2} (f(z)-1)\right] \hat z $$

First of all, let's grab what parts we've already evaluated.

$$ \vec B(0) = \frac{\mu_0 N I}{\sqrt{(L/2)^2+R^2}} \hat z$$
$$ \vec B(L/2) = \frac{\mu_0 N I}{\sqrt{L^2+R^2}} \hat z$$

And from this we can easily find the result for f(z). 

$$ f(0) = \frac{L}{\sqrt{(L/2)^2+R^2}} $$
$$ f(L/2) = \frac{L}{\sqrt{L^2+R^2}} $$

As we know $B_0$ is the magnetic field without the magnetizable material, so naturally we seek to find how it differs from that value. So let's state the f values in terms of the B ones.

$$ f(0) = B_0 \frac{L}{\mu_0 N I} $$
$$ f(L/2) = B_0 \frac{L}{\mu_0 N I} $$

Turns out they're hte same at both points (that is, the same scaling). Thus, we can rewrite the relation as...

$$ \vec B(z) \approx (1+\chi_m)\left[ B_0(z) + \mu_0 \frac{NI}{L} \frac{\chi_m}{\chi_m+2} (B_0\frac{L}{\mu_0 N I}-1)\right] \hat z $$

Which we can simplify a bit

$$ \Rightarrow \vec B(z) \approx (1+\chi_m)\left[ B_0(z) +  \frac{\chi_m}{\chi_m+2} (B_0-\mu_0 \frac{NI}{L})\right] \hat z $$

$$ \Rightarrow \vec B(z) \approx (1+\chi_m)\left[ B_0(z) +  \frac{\chi_m}{\chi_m+2} (B_0-\frac{B_0}{f(z)})\right] \hat z $$

$$ \Rightarrow \vec B(z) \approx B_0 (1+\chi_m)\left[ 1 +  \frac{\chi_m}{\chi_m+2} (1-\frac{1}{f(z)})\right] \hat z $$

Thus the amplification itself (at the ends) is

$$ (1+\chi_m)\left[ 1 +  \frac{\chi_m}{\chi_m+2} (1-\frac{1}{f(z)})\right] $$

with

$$ f(0) = \frac{L}{\sqrt{(L/2)^2+R^2}} $$
$$ f(\pm L/2) = \frac{L}{\sqrt{L^2+R^2}} $$

And that is our answer. 

As for when $\chi_m$ gets large, well, that becomes...

$$ \vec B(z) \approx B_0 \chi_m \left[ 2 - \frac{1}{f(z)}\right] \hat z $$


Here the L>>R simplification will actually be useufl since this can simplify down to something understandable. 

$$ f(0) \approx 1/2 $$
$$ f(\pm L/2) \approx 1 $$

$$ \vec B(0) \approx 0 $$
$$ \vec B(\pm L/2) \approx B_0 \chi_m \hat z $$

Which is to say the field at the ends magnifies by a factor about equal too $\chi$ and then drops to near nothing in the middle.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

So this is similar to the moon problem. First of all, for a point dipole there is no magnetization so H and B are effectively the same. So we can find H due to the dipole alone just by removing the factor of $\mu_0$.

$$ \vec H_d = \frac{1}{4\pi} \left[ \frac{3(\vec m \cdot \vec r_2)\hat r_2 - \vec m}{r_2^3} \right] $$

Since our sphere is arbitrary and symmetric we can just declare the magnetic moment to be pointing along the axis, that is, $\vec m = m\hat z$. Also $r_2$ is the only r so we can just replace it with r.

$$ \Rightarrow \vec H_d = \frac{1}{4\pi} \left[ \frac{3mr_z\hat r - m \hat z}{r^3} \right] $$

And for clarity the magnetic field of the dipole alone is

$$ \Rightarrow \vec B_d = \frac{\mu_0}{4\pi} \left[ \frac{3mr_z\hat r - m \hat z}{r^3} \right] $$

We can (and probably should) write the fields in spherical coordinates

$$ \vec H_d = \frac{1}{4\pi r^3} \left[ 2cos\theta \hat r + sin\theta \hat \theta \right] $$

$$ \vec B_d = \frac{\mu_0}{4\pi r^3} \left[ 2cos\theta \hat r + sin\theta \hat \theta  \right] $$

Now we ultimately want the H-field of the sphere itslef, both inside and outside, and then we'll add it to the dipole's H-field to get the total. 

We can get the magnetization of the sphere relatively easily, as $\vec M = (\mu/\chi) \vec B$. However, we are only given $\mu$, so let's make sure to declare what $\chi$ is in terms of that.

$$ \mu = \mu_0(1+\chi) \rightarrow \chi = \frac{\mu-\mu_0}{\mu_0} $$

We will continue writint it as $\chi$ since that is easier to look at, though. Anyway, this gives us the magnetization of the sphere.

$$ \vec M_s = \frac{\mu_0\mu}{4\pi \chi r^3} \left[ 2cos\theta \hat r + sin\theta \hat \theta  \right] $$

Now, this is very similar to the moon problem, except for the fact that there is no "inner boundary" and the sphere is just laid on top of the dipole. Realistic? No. Easier to calculate? Yes. The magnetization form is identical between the two situations, so we know that the divergence is zero and there is no volume magnetic charge. There is also no inner surface. There is, however, an outer surface, and that surface's "fake magnetic charge" is

$$ \sigma = (\vec M \cdot \hat n) = \frac{\mu_0 \mu m}{2 \pi \chi R^3}cos\theta $$

This is the charge distribution that defines the sphere's H-field, both inside and out. We just need to solve it like an electrostatics problem, and it just so happens we have spherical harmonics that can help us with that. The magnetic potential has to match at the boundaries, and so both sides need to be of the cos$\theta$ spherical harmonic. Region I is inside, Region II is outside, giving us:

$$ \psi_I = Arcos\theta $$
$$ \psi_{II} = \frac{B}{r^2}cos\theta $$

There is no "middle" area, which is really good for us as that simplifies things considerably. At r=R, the need to be equal:

$$ AR = B/R^2 \Rightarrow AR^3 = B$$

And their derivatives need to have a differnece of the fake charge at the boundary.

$$ A + 2B/R^3 = \frac{\mu_0 \mu m}{2 \pi \chi R^3} $$

Note that every instance of cos$\theta$ has been suppressed as it is in every term and can be divided out. 

Anyway, two equations, two unknowns. Substitute $B=AR^3$.

$$ 3A = \frac{\mu_0 \mu m}{2 \pi \chi R^3} $$
$$ \Rightarrow A = \frac{\mu_0 \mu m}{6 \pi \chi R^3} $$

And thus B is...

$$ B = \frac{\mu_0 \mu m}{6 \pi \chi} $$

And so our potentials are

$$ \psi_I = \frac{\mu_0 \mu m r}{6 \pi \chi R^3}cos\theta $$
$$ \psi_{II} = \frac{\mu_0 \mu m}{6 \pi \chi r^2}cos\theta $$

From here, we can get the H-field due to the sphere, since $\vec H = -\nabla \psi$

$$ \vec H_{sI} = \frac{\mu_0 \mu m}{6 \pi \chi R^3}cos\theta \hat r - \frac{\mu_0 \mu m}{6 \pi \chi R^3}sin\theta \hat \theta $$
$$ \Rightarrow \vec H_{sI} = \frac{\mu_0 \mu m}{6 \pi \chi R^3}(cos\theta \hat r - sin\theta \hat \theta) $$

$$ \vec H_{sII} = - \frac{\mu_0 \mu m}{3 \pi \chi r^3}cos\theta \hat r - \frac{\mu_0 \mu m}{6 \pi \chi r^3}sin\theta \hat \theta $$
$$ \Rightarrow \vec H_{sII} = - \frac{\mu_0 \mu m}{6 \pi \chi r^3}(2cos\theta \hat r + sin\theta \hat \theta) $$

We are naturally wary of the $\mu_0$ term in an H-field, but we also note that $\mu$ and $\chi$ are present, so it's all right. 

However, unlike the moon, the interior dipole is still present here. So the full field will have us add the two together in each region. Thus, our final answer would be...

$$ \vec H_{I} = \frac{\mu_0 \mu m}{12 \pi \chi R^3}(cos\theta \hat r - sin\theta \hat \theta) + \frac{1}{4\pi r^3} \left[ 2cos\theta \hat r + sin\theta \hat \theta \right]$$

$$ \vec H_{II} = - \frac{\mu_0 \mu m}{12 \pi \chi r^3}(2cos\theta \hat r + sin\theta \hat \theta) + \frac{1}{4\pi r^3} \left[ 2cos\theta \hat r + sin\theta \hat \theta \right]$$

While yes we could combine the like terms for $\hat r$ and $\hat \theta$, we think this form is more illustrative of what's going on, as the contribution of the sphere and the dipole are clearly separate. The dipole's contribution is unchanging with r, meanwhile the sphere's contribution changes depending on if we are inside or outside. We also note that while the inteiror sphere field is not *constant*, it doeds not depend on r, which matches with what we know from electric fields.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

So there are two forces acting on the loop. One is gravity, with a force of mg. Thus, we need to be pulling back on the loop also with a force of mg. The total mass is $2\pi a \rho$, that is the length of the loop times the mass of the loop, so the actual force we need to replicate is $-2\pi a \rho g$. The minus sign is there since g is negative and we need to pull agianst it. 

We know, somewhat, the force on a current loop:

$$ \vec F = \int \vec J(\vec r) \times \vec B(\vec r) d^3r $$

The problem is that this does not provide us a unique solution. Sure, as J is the current density which we are told is steady, so the current only exists in a small area. However we can imagine several field distributions that would provide the same force, so we are once again at a bit of a standstill.

However, one thing we do know is that the system is azimuthally symmetric, that is, every point on the ring feels the same field. This would result in no force at all if the ring was just sitting there, but gravity is pulling on it, and if it *were* to move down, it would feel a gradient, and thus a force. 

Now, in order to get this azimuthal symmetry (which is required for a levitating and not spinning ring) the magnetic field and magnetization have to be azimuthally symmetric as well. We shall assume that, since this is a permanent magnet, that its magnetization is constant all the way through. For now, let's just call this M and say it's pointing in the z direction. (We will worry about sign later). The plan is to start with a linear M, find the magnetic field, and then find the force that magnetic field actually exierts on the loop, and then solve for the actual value of M. 

So, we have a magnetization. Trying to find the magnetic field (B) outside of the bar magnet from this is going to be somewhat difficult, so isntead let us seek the H-field using it's "pseudo-electromagnetic" properties.

$$ \vec B = \mu_0 M \hat z + \mu_0 \vec H $$

H is generated by fake magnetic charges given by $-\nabla \cdot \vec M$. The divergence does not exist within the magnet or without, but on the edges we do have $ \sigma_M = \vec M \cdot \hat n$.  Since the arrows do not "point" at the curved edge, only the flat edges will have any fake surface charge. There will be surface charge at z=h and at some distance really far away, let's call it z=L where L approaches infinity. Thus, we can find H if we can find E for two oppositely charged discs. The thing is we'd like to know the effect everywhere rather than just a single point, since we need to have the force *not* acting on the center of the loop, but the line of the loop itself.

Anyway, our surface charges are $\pm M\hat z$, where the minus is associated with the lower disc. (the dot procuct itself has no minus sign, M by defualt points up, but the normal points down, so negative on the bottom.) The influence of the upper disc, being at infinity, is going to be essentially nonexistent.  

Now, under normal circumstances we would use the electric field of a disc. But we don't have that. however, what we do have is a small radius r, which means we can treat the loop as "far away" from the disc, which effectively makes it a "point charge" as far as the loop is concerned. How much "charge" is in here? well, since it's a constant "surface charge" we just multiply by the area to get:

$$ - M \pi r^2 $$

Now the electric field of a point charge is $\frac{Q}{4\pi \epsilon_0 r^2} \hat r$. Now, for a magnetic adaptation of this, we replace $\epsilon_0$ with 1 (since we are in H field rather than B field). Since we're already using r let's replace this radius with another variable, let it be R. Thus...

$$ \vec H = \frac{-M\pi r^2}{4\pi R^2} \hat R = \frac{-M r^2}{4 R^2} \hat R $$

A purely radial H-field makes sense when we're treating it as a "point" charge. 

Now, the great thing? Outside the bar magnet the H-field is all there is! So...

$$ \vec B = -\frac{\mu_0 M r^2}{4R^2}\hat R $$

So now that we have B we need the force on the loop.

$$\vec F = \int \vec J(\vec r) \times \vec B(\vec r) d^3r$$

Our current density is constant and we just integrate around it. At a specific spot on the loop, B is also constant. 

$$\vec F = \int \vec J \times \vec B d^3r$$

Technically this is zero since in equilibrium neither are changing. However, if we let the loop *fall*, we can get a force from a "slight adjustment" which would be (from the notes)...

$$\vec F = \int \vec J \times [(\vec r \cdot \nabla)\vec B] d^3r $$

In our specific case, this will be applied somewhat unusually. 

That r-vector is essentially an arbitrary vector pointing from some origin. We choose the origin to be the tip of the magnet, so the r-vector is purely radial and of a length $\sqrt{h^2+a^2}$. This makes $\vec r \cdot \nabla = \sqrt{h^2+a^2}\frac{\partial}{\partial R}$, since, once again, the radius of the magnet itself is little r so we use big R as the variable. Since hte gradient is now a simple derivative, we can actually take it.

$$\vec F = \int \vec J \times \sqrt{h^2+a^2} \frac{-\mu_0 M r^2}{2R^3} \hat R d^3r $$

Now J is purely in the azimuthal direction, so we have a cross product from azimuthal to radial. One might think this means we go directly to another unit vector, but we have to remember we're modeling this in spherical.

$$\Rightarrow \vec F = \int \sqrt{h^2+a^2} \frac{-J\mu_0 M r^2}{2R^3} \hat \phi \times \hat R d^3r $$

The actual direction here can be deduced. The current flows along the loop in the phi direction, and the radial always points away from the magnet right through the current loop. The effect is that the direction is efectively "reflected" across the central axis, making the resulting arrow point "inward" and "downward" along the loop. We note that when we integrate along the loop, the "inward" components will all vanish, leaving only the directly "downward" facing components. (which is what we expect). Since we have a resulting downward component, it seems like a good idea to represent the answer in cylindrical coordinates in terms of the z direction and ignore the other directions. 

Thus, we need to throw in a cosine term to convert to just that component. This cosine ends up being $\frac{h}{\sqrt{h^2+a^2}}$. Thus...

$$\Rightarrow \vec F = \int h \frac{J\mu_0 M r^2}{2R^3} \hat z d^3r $$

Where the sign flips because the vector component we found was downward. 

Now we need to be careful with this integral. Remember that J = $I\delta(z)\delta(\rho-a)$ NOT just the physical length of the loop. Which means that when we convert to a length, we end up with...

$$\Rightarrow \vec F = \int h \frac{I\mu_0 M r^2}{2R^3} \hat z dl $$

Anyway, big R is now a constant since we are only on the loop. In cylindrical, we replace dl with $R d\phi$, with, agian, R being constant.

$$\Rightarrow \vec F = \int_0^{2\pi} h \frac{I\mu_0 M r^2}{2R^2} \hat z d\phi $$
$$\Rightarrow \vec F = h \frac{2\pi I\mu_0 M r^2}{2R^2} \hat z $$
$$\Rightarrow \vec F = h \frac{\pi I\mu_0 M r^2}{(h^2+a^2)} \hat z $$

Now set force to mg and then solve for M.

$$ M = \frac{(2\pi a \rho g)(h^2+a^2)}{\pi hI\mu_0 r^2} $$
$$ \Rightarrow M = \frac{2 a \rho g(h^2+a^2)}{hI\mu_0 r^2} $$

And so in vector form...

$$\vec M = \frac{2 a \rho g(h^2+a^2)}{hI\mu_0 r^2} \hat z$$

Now the units of M should be A/m, and this does in fact have those units, so we submit it as our answer.
