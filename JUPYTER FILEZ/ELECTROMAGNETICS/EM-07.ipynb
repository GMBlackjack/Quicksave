# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

First of all, make the two lines electromagnetic mirrors. This will create two mirror charges... but also each mirror will mirror the other mirror, meaning we ALSO get a dobule mirrored charge on the exact opposite side, meaning the mirrored system is a system of four charges of alternating sign. 

Which makes sense becuase to get a "+" shape you'd definitely need four charges anyway. We can draw up a quick diagram:



Blue is one sign of charge, red is the other. The charge itself of the original point was not specified so each could be either. We have specified angle and d here, but it could be completely arbitrary--it's just the overall arrangement that would be the same.

All the charges must have the same relative value, so we assign it $\pm Q$ and let the charge we care about in the first quadrant be +Q. Each point charge adds to the potential based on the x and y coordinates of everything. The arbitrariness (and the fact that none of the charges are origin-centered) does interfere with this, but we can work it out. Cartesian seems the best for this. Each point charge will identically be written as:

$$ \varphi = \frac{Q}{4\pi \epsilon_0} \frac{1}{|\vec r - \vec r'|} = \frac{Q}{4\pi \epsilon_0} \frac{1}{\sqrt{(x-x')^2 + (y-y')^2 + z^2}} $$

Note the presence of z. We have plates, this is not a 2D problem. 

Now, the absolute value of the vectors is determined by our x and y coordinates we're testing and the position of the particle itself. x and y are the coordinates we test, while the x' and y' are the coordinates of the points themselves. Thus, we end up with a total potential of everything added together:

$$ \varphi = \frac{Q}{4\pi \epsilon_0} \left(  \frac{1}{\sqrt{(x-dcos\beta)^2 + (y-dsin\beta)^2  + z^2}} -  \frac{1}{\sqrt{(x+dcos\beta)^2 + (y-dsin\beta)^2  + z^2}} \\-  \frac{1}{\sqrt{(x-dcos\beta)^2 + (y+dsin\beta)^2  + z^2}} +  \frac{1}{\sqrt{(x+dcos\beta)^2 + (y+dcos\beta)^2  + z^2}} \right) $$

Anyway, that's the potential due to the four-charge system. At (0,0) it naturally goes to zero as everything cancels. It also cancels when just x or just y equals zero, so it's all good. 

Now, to find the charge distribution, we return to the following relation:

$$ \sigma = \epsilon_0 \hat z \cdot \vec E(z=0) = -\epsilon_0 \left| \frac{\partial \varphi}{\partial z} \right|_{z=0} $$

Which can be converted into y and x terms easily enough.

$$ \sigma_x = -\epsilon_0 \left| \frac{\partial \varphi}{\partial x} \right|_{x=0} $$
$$ \sigma_y = -\epsilon_0 \left| \frac{\partial \varphi}{\partial y} \right|_{y=0} $$

The general form of the derivative is

$$ \frac{\partial}{\partial x}  \frac{1}{\sqrt{(x-a)^2 + (y-b)^2 + z^2}} = \frac{a-x}{|(x-a)^2 + (y-b)^2 + z^2|^{3/2}} $$
$$ \frac{\partial}{\partial y}  \frac{1}{\sqrt{(x-a)^2 + (y-b)^2 + z^2}} = \frac{b-y}{|(x-a)^2 + (y-b)^2 + z^2|^{3/2}} $$

Where a and b replace the trig functions so it's easy to write. Now, if we evaluate at x=0 and y=0 respectively:

$$ = \frac{a}{|(a^2 + (y-b)^2 + z^2|^{3/2}} $$
$$ = \frac{b}{|(x-a)^2 + b^2 + z^2|^{3/2}} $$

Which makes sense as the charge density on a plate on the x=0 plane would have charge density defined by y and z, and x and z for the other plate on the y=0 plane. 

So, we can start writing down our answers now.

$$ \sigma_x = -\frac{Q}{4\pi} \left( \frac{a}{|(a^2 + (y-b)^2 + z^2|^{3/2}} - \frac{-a}{|(a^2 + (y-b)^2 + z^2|^{3/2}} \\- \frac{a}{|(a^2 + (y+b)^2 + z^2|^{3/2}} + \frac{-a}{|(a^2 + (y+b)^2 + z^2|^{3/2}} \right) $$

$$ = -\frac{Q}{4\pi} \left( \frac{2a}{|(a^2 + (y-b)^2 + z^2|^{3/2}} - \frac{2a}{|(a^2 + (y+b)^2 + z^2|^{3/2}} \right) $$
$$ = -\frac{Qa}{2\pi} \left( \frac{1}{|(a^2 + (y-b)^2 + z^2|^{3/2}} - \frac{1}{|(a^2 + (y+b)^2 + z^2|^{3/2}} \right) $$

There does not appear to be a clear way to simplfy this further, so we substitute back in.

$$ \sigma_x = -\frac{Qdcos\beta}{2\pi} \left( \frac{1}{|((dcos\beta)^2 + (y-dsin\beta)^2 + z^2|^{3/2}} - \frac{1}{|((dcos\beta)^2 + (y+dsin\beta)^2 + z^2|^{3/2}} \right) $$

The y plane is similar, just flipped. 

$$ \sigma_y = -\frac{Qdsin\beta}{2\pi} \left( \frac{1}{|((dsin\beta)^2 + (x-dcos\beta)^2 + z^2|^{3/2}} - \frac{1}{|((dsin\beta)^2 + (x+dcos\beta)^2 + z^2|^{3/2}} \right) $$

Note that at x=0 or y=0 the subtraction makes the charge density go to zero. This matches what we know about corners that are angles less than 180 degrees: charge does not like congregating in them. Now, if we had placed the charge on the *other* side, we would have ended up with a point situation, which could go very differently depending on how we chose to formulate it, but we would expect an explosion of charge at the point.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

Right, so, method of images has partially solved this question for us already. we can replace the grounded sphere with two interior charges, each one based on one of the other charges indepedently. The one charge situation has already been solved. If all the charges are on a line (let it be the z axis), then we have solutions of:

$$ b = R^2/s$$
$$ q' = -Rq/s $$

s is the distance of the original charge from the sphere's center, b is the distance of the new charge from the spheres center, q is the old charge, q' is the new charge, and R is just the radius of the sphere. 

If our original charges are Q and Q', let's add a subscript to help differentiate. Original, or exterior charges, are $Q_o, Q_o'$, and the interior charges can be $Q_i, Q_i'$. 

We can now make a sort of "table" of charge properties:

$$ name ; charge ; position $$
$$ Q_o ; Q_o ; 2R $$
$$ Q_o' : Q_o' ; -5R $$
$$ Q_i ; -Q_o/2 ; R/2 $$
$$ Q_i' ; -Q_o'/5 ; -R/5 $$


So we now ask what the force is on charge $Q_o'$ in order to find X. Fortunately, every last one of these are point charges, so we can just use the electric field of a pount charge and find what it is for all of them at -5R. The total distances are then 5R - R/5, 5R + R/2, and 5R+2R, or 24R/5, 11R/2, 7R. 

$$ F = \frac{Q_o'}{4\pi\epsilon_0} \left( - \frac{Q_o}{49R^2} + \frac{2Q_o}{121R^2} + \frac{5Q_o'}{576R^2} \right)$$

As we can see, all whole numbers, but definitely not going to be neat. As we were told. Signs chosen since the negative direction is "repulsive" in the position we placed our charge. Anyway, as all charges are positive and we only care about the *sign* of F, the leading coefficients and leading $Q_0'$ are irrelevant here. 

$$ F \propto - \frac{Q_o}{49R^2} + \frac{2Q_o}{121R^2} + \frac{5Q_o'}{576R^2} $$

The $R^2$ in every term is also irrelevant. Simplify as well. 

$$ F \propto - \frac{Q_o}{49} + \frac{2Q_o}{121} + \frac{5Q_o'}{576}$$

Now, if $Q_o' = 0$ we get a negative result, so at the very least for a small charge it's a negative, which means repulsive force. (As the charge is positioned on the negative side). And if we make it very LARGE, it dominates, making a positive (attractive) force. So now what we're looking for is the middle point, where F=0, and to solve for the ratio.

$$ 0 = - \frac{Q_o}{49} + \frac{2Q_o}{121} + \frac{5Q_o'}{576}$$
$$ \Rightarrow 0 = \left( \frac{2}{121} - \frac{1}{49} \right)Q_o + \frac{5Q_o'}{576}$$
$$ \Rightarrow - \left( \frac{2}{121} - \frac{1}{49} \right)Q_o = \frac{5Q_o'}{576}$$
$$ \Rightarrow - \frac{576}{5}\left( \frac{2}{121} - \frac{1}{49} \right)Q_o = Q_o'$$
$$ \Rightarrow -  \frac{576}{5}\left( - \frac{23}{5929} \right)Q_o = Q_o'$$
$$ \Rightarrow \frac{13248}{29645} Q_o = Q_o'$$

At which point we just had a computer simplify the fraction. 

...computer says it is in simplified terms. Good gravy, that is ugly. It is close to 1/2, 0.447.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

This is a Green's funciton problem. Fortunately, we already have the Green's function for above a z=0 plane in the notes. 

$$ G_D(\vec r_1 , \vec r_2) = \frac{1}{4\pi\epsilon_0}\left[ \frac{1}{\sqrt{(x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2}} - \frac{1}{\sqrt{(x_1-x_2)^2+(y_1-y_2)^2+(z_1+z_2)^2}} \right] $$

With this, we can calculate the potential using the Dirchlet Green's Funciton Relation. (Dirchlet because the potential is specified, and it was an assumption that went into making the above Green's Function.)

$$ \varphi(\vec r_2) = \int_V G_D(\vec r_1 , \vec r_2) \rho(\vec r_1) d^3 r_1 - \epsilon_0 \int_S \varphi(\vec r_1) \nabla_1 G_D(\vec r_1 , \vec r_2) \cdot dS_1 $$

Now the second term in this equation is obviously giong to become our integral over the surface $S_0$. But what about the first term? We know G, but do we know $\rho$? Well, fortunately for us, in this case $\rho$ is just the charge density in the volume we care about. We are specifically asked for the potential *away* from the plane (and we arguably already know it on the plane) so the volume charge density... is nothing! All the charge MUST be on the plane. So the first term is zero!

Let's examine the second term now.

$$ \varphi(\vec r_2) = - \epsilon_0 \int_S \varphi(\vec r_1) \nabla_1 G_D(\vec r_1 , \vec r_2) \cdot d\vec S_1 $$

First of all, $\varphi$ is our potential, and it is eitehr zero or constant, which allows us to simplify and change the bounds of integration.

$$ \varphi(\vec r_2) = - \epsilon_0 \varphi_0 \int_{S_0}  \nabla_1 G_D(\vec r_1 , \vec r_2) \cdot d\vec S_1 $$

If we pull the constant term out of the G function, we have:

$$ \varphi(\vec r_2) = - \frac{\varphi_0}{4\pi} \int_{S_0}  \nabla_1 \left[ \frac{1}{\sqrt{(x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2}} - \frac{1}{\sqrt{(x_1-x_2)^2+(y_1-y_2)^2+(z_1+z_2)^2}} \right] \cdot d^2 \vec S_1 $$

In essence, it all depends on the gradient of the G function. Well, all right, let's see what that gradient equals with respect to $r_1$.

$$  \left[ \frac{x_2-x_1}{((x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2)^{3/2}} - \frac{x_2-x_1}{((x_1-x_2)^2+(y_1-y_2)^2+(z_1+z_2)^2)^{3/2}} \right] \hat x \\
+  \left[ \frac{y_2-y_1}{((x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2)^{3/2}} - \frac{y_2-y_1}{((x_1-x_2)^2+(y_1-y_2)^2+(z_1+z_2)^2)^{3/2}} \right]\hat y \\
+  \left[ \frac{z_2-z_1}{((x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2)^{3/2}} - \frac{-z_2-z_1}{((x_1-x_2)^2+(y_1-y_2)^2+(z_1+z_2)^2)^{3/2}} \right]\hat z $$

Now tehre's a clear pattern here where the numerator is not different between the two terms, except in the case of the *last* coordinate, z. Fortunately we are integrating over the surface, which means we know that, for sure, $z_1=0$. This simplifies things CONSIDERABLY to:

$$ 0 \hat x + 0 \hat y + \left[ \frac{2z_2}{((x_1-x_2)^2+(y_1-y_2)^2+z_2^2)^{3/2}} \right]\hat z $$

The denominator is the cube of a vector, so we can rewrite it

$$ \left[ \frac{2z_2}{|\vec r_1 - \vec r_2|^3} \right]\hat z = \left[ \frac{2z_2}{|\vec r_2 - \vec r_1|^3} \right]\hat z $$

We can set the subtration any direction we want as it is an absolute value. We insert it back into our Green's relation...

$$ \varphi(\vec r_2) = - \frac{\varphi_0}{4\pi} \int_{S_0}  \left[ \frac{2z_2}{|\vec r_2 - \vec r_1|^3} \right]\hat z  \cdot d\vec S_1 $$

Have to be careful with this dot product. It's the normal to the surface.

$$ \varphi(\vec r_2) = - \frac{\varphi_0}{4\pi} \int_{S_0}  \left[ \frac{2z_2}{|\vec r_2 - \vec r_1|^3} \right]\hat z  \cdot \vec n d^2 r_1 $$

Now since we are in the positive half, that means the normal points "out" or DOWN, thus flipping the sign when we take the dot product to get a flat number.

$$ \Rightarrow \varphi(\vec r_2) = \frac{\varphi_0}{4\pi} \int_{S_0}  \frac{2z_2}{|\vec r_2 - \vec r_1|^3} d^2  r_1 $$

As we are integrating over $r_1$, $z_2$ is a constant and can come out.

$$ \Rightarrow \varphi(\vec r_2) = \frac{\varphi_0z_2}{2\pi} \int_{S_0}  \frac{1}{|\vec r_2 - \vec r_1|^3} d^2  r_1 $$

And if we declare 1 as prime and 2 as our variable, we get...

$$ \Rightarrow \varphi(\vec r) = \frac{\varphi_0z}{2\pi} \int_{S_0}  \frac{d^2  r'}{|\vec r - \vec r'|^3}  $$

Which is for above the plate, naturally if we included below it would be

$$ \Rightarrow \varphi(\vec r) = \frac{\varphi_0|z|}{2\pi} \int_{S_0}  \frac{d^2  r'}{|\vec r - \vec r'|^3}  $$

Which is what we sought.
