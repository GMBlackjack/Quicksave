# Problem 1 \[Back to [top](#toc)\]

First of all, since we're asked for what is basically an approximation we can make a highly simplified model. Specifically, we can make a *square* wire...



In our approximation, we say that everywhere except in the corner, the electric field must be more or less constant since the current also has to be more or less constant. This effectively means that the long "extensions" of the wire can be thought of as (very long) parallel plate capacitors. We declare the flow to be coming in from the left and then going down. If we ignore the corner part, we can declare that just before the current enters the corner, it is at a constant potential and as it exits, it is also at a (slightly lower) constant potential. Let's call these $V_1$ and $V_2$ which are more or less arbitrary, just that the first must be at a higher potential than the second. 

In this case, the corner becomes a cube, a cube that we can assign boundary conditions to. The -x side has $V_1$, the -y side has $V_2$, and every other side is restricted by not permitting the electric field or current to flow out of the system, which restricts the derivative. While the boundaries in the z direction are somewhat irrelevant given the symmetry of the problem and direction of flow, we shall still keep them in mind. 

A reminder: the current has to flow precisely perpendicular to the boundaries where the field cannot leave the wire.



Here is the cross-sectional area that we are using. 

Let's try a simplification. Imagine, if we will, the problem in only one axis of motion. A current (or charge or what have you) has to be stopped moving in the direction it originaly comes in (and ejected out another direction). We can model this  as a stream approaching a single point charge. Since we are in a material, the motion of a charge dissapears when there is no electric field, so what we need to do is cancel out the constant electric field. If we do this with a point charge at a position in the center of a "cube" like the above model, we have a radius of L/2 for the point charge's effect, and its field. Which is to say...

$$ \frac{Q}{4\pi\epsilon_0 (L/2)^2} = E$$

Technically speaking there should be a minus sign somewhere but we only care about the magnitude at the moment. Simplifying a bit...

$$ \Rightarrow \frac{Q}{\pi\epsilon_0 L^22} = E$$
$$ \Rightarrow Q = E \pi \epsilon_0 L^2  $$

Now, something fun here is that L^2 happens to be the cross sectional area of our cube.

$$ \Rightarrow Q = E \pi \epsilon_0 A $$

This is actually very important... but it won't be easy to figure out exactly how, yet. So let's back up a bit and look at the problem a different way to get hte insight. The point charge we've placed is fighting agianst something, that something is a potential. 

$$ V = ED $$

Where D is the length of the wire. It'll cancel out so we won't dwell on it. As we are in an Ohmic material, V=IR, and R is given by D/A$\sigma$

$$ \Rightarrow I \frac{D}{A\sigma} = ED$$ 

Now the D cancels, and we can solve for... EA.

$$ \Rightarrow \frac{I}{\sigma} = EA$$ 

So now we have the insight needed. With the substitution, we have:

$$ \Rightarrow Q = \pi \epsilon_0 \frac{I}{\sigma}  $$

And this makes a good estimate. Technically speaking, we should multiply by 2 as we also need another charge to push the incoming current away once it arrives, but a factor of 2 does not alter a very vague order of magnitude estimate. Ignoring the factor constants, the result is...

$$ \Rightarrow Q \approx \epsilon_0 \frac{I}{\sigma}  $$

Which is exactly what we sought. 

Form the last one we have $Q = \epsilon_0 I / \sigma$. 

My laptop power supply says 19.5 V and 11.8 A, so that's what I'll use. As for the conductivity... copper is pretty commonly used in wires, what's its conductivity? (We find this later as there was units confusion) Naturally $\epsilon_0$ is 8.854e-12. 

First we shoudl check units. Current is in amps, or coulombs per second. The permativity is $\frac{s^2C^2}{m^3kg}$. It took a while to track down what conductivity was, actually, but the SI unit is $ \frac{sC^2}{kgm^3} $. Which meanst hat, yes, the result of all the units together means Q is given in C, Coulombs, as it should be.

However we need to be careful about the number we're using for conductivity! What units is it in? The SI Units were S/m, but it's apparently often given in $\mu$S/m (where S is Siemens). Well after much digging we found that a Siemen is just an inverted Ohm, and we can just invert the resistivity. For copper resistivity is 1.68e-8$\Omega$m and so conductivity is 5.952e7 S/m. 

Which will give us a grand total charge of... 

$$ 1.755e-18 C $$

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

So, in a sense, we can think of the sphere itself, since it is a perfect conductor, as something at constant potentail V and the infinitely deep point as something at potential 0. 

Now, since we are finding resistance, we assume constant current. This does NOT mean constant current DENSITY since we are dealing with a spherical situation--the current is going to flow in every direction on a hemisphere and will naturally lose density as it goes out, while the total current over the entire hemisphere will remain the same. That said, we can still say that at the interfaces between two substances of different conductivity the current density has to be the same.

Also we can't use Gauss' law since it's not the *entire* sphere we're dealing with, just half of it. 

One thing we can use $\vec J = \sigma \vec E$. So the density of our current is very obviously giong to go like $1/r^2$. We can still use this at the boundary to force a condition: at r=b, we have:

$$ \sigma_2 E_2 = \sigma_E E_E $$

The question now is what the potential drop is at the boundary. We now need an expression related to the voltage V so we can solve and get a solution. Let the potential of the sphere be V and the potential at a distance be 0. The basic form of this expression would be:

$$ V = \int_b^a E_2 dr + \int_\infty^b E_E dr $$

The question now is what those individual potensials are. To do this, let's play around with some current.

$$ J = I / A $$

A is the area of a hemisphere so $A = 2\pi r^2$ which gives us $J = I/2\pi r^2$. 

$$ J = \sigma E = -\sigma \frac{d\varphi}{dr}$$

Re-arrange to solve for the potential derivative.

$$\frac{d\varphi}{dr} = -\frac{J}{\sigma} = -\frac{I}{\sigma2 \pi r^2}$$

Integrate with respect to r...

$$\int \frac{d\varphi}{dr} dr = -\int \frac{I}{\sigma2 \pi r^2} dr$$

$$ \left[ \varphi \right]_? = \left[ \frac{I}{\sigma 2 \pi r} \right]_?  $$

With this we can now evaluate the potential difference for b to a and infinity to b. 

$$ \left[ \varphi \right]_b^a = \left[ \frac{I}{\sigma_2 2 \pi r} \right]_b^a = \frac{I}{2 \sigma_2 \pi}\left( \frac{1}{a} - \frac{1}{b} \right)  $$
$$ \left[ \varphi \right]_\infty^b = \left[ \frac{I}{\sigma_E 2 \pi r} \right]_\infty^b = \frac{I}{2 \sigma_E \pi}\left( \frac{1}{b} \right)  $$

And so...

$$ V = \frac{I}{2 \sigma_2 \pi}\left( \frac{1}{a} - \frac{I}{b} \right) + \frac{1}{2 \sigma_E \pi}\left( \frac{1}{b} \right) $$

Becasue we have a way to find it in clsoed form, we can use V=IR directly. Which just means R = V/I

$$ R = \frac{1}{2 \sigma_2 \pi}\left( \frac{1}{a} - \frac{1}{b} \right) + \frac{1}{2 \sigma_E \pi}\left( \frac{1}{b} \right) $$

We can use the same relation as the previous part, but we have to treat $\sigma$ as a function.

$$ \left[ \varphi \right]_\infty^a = V = -\int_\infty^a \frac{I}{\sigma(r)2 \pi r^2} dr  $$

$$  V = -\int_\infty^a \frac{I (1 - (r/L)^2 \beta e^{-r/L})}{\sigma_0 2 \pi r^2} dr $$
$$  = - \frac{I}{\sigma_0 2 \pi}\int_\infty^a \frac{1 - (r/L)^2 \beta e^{-r/L}}{r^2} dr $$
$$  = - \frac{I}{\sigma_0 2 \pi}\int_\infty^a \frac{1}{r^2} - \frac{(r/L)^2 \beta e^{-r/L}}{r^2} dr $$
$$  = - \frac{I}{\sigma_0 2 \pi}\int_\infty^a \frac{1}{r^2} - \frac{\beta e^{-r/L}}{L^2} dr $$

$$  = - \frac{I}{\sigma_0 2 \pi}\left[ -\frac{1}{r} + \frac{\beta e^{-r/L}}{L^3} \right]_\infty^a $$
$$  = \frac{I}{\sigma_0 2 \pi}\left( \frac{1}{a} - \frac{\beta e^{-a/L}}{L^3} \right)$$

Divide by I to get R.

$$ R = \frac{1}{\sigma_0 2 \pi}\left( \frac{1}{a} - \frac{\beta e^{-a/L}}{L^3} \right) $$

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

Despite being, evidently, a Biot-Savart problem, this is actually revealted to be a superposition problem. So let's start with the easeir problem: that is, if the width "b" is infinitesimal and the plate might as well be a wire. Just let the current flowing through it be I. This is a well known result: field rotates counterclockwise in perfect circles. We can find the precise formulation in equation 10.29

$$\vec B = \frac{\mu_0 I}{2\pi \rho} \hat \phi $$

In the case of our two points positioned at (a,0) and (0,a) we have:

$$\vec B(0,a) = - \frac{\mu_0 I}{2\pi a} \hat x $$
$$ \vec B(a,0) = \frac{\mu_0 I}{2\pi a} \hat y $$

Now we might see immediately that the (a,0), that is the x version, has a bennefit: the plate can be thought of as a bunch of wires all lined up in a row pointing directly at the point. The superposition integral here won't have any unusual angles at all, and *every single magnetic field can only provide a push in the +y direction no matter where it is on this line.* The only change we have to worry about is the change in $\rho$ and the change in I, since I is the total current, and infinitesimal current is a bit different. 

In the end, we get this:

$$ \vec B = \int_{-b/2}^{b/2} \frac{\mu_0 I}{2\pi (a-x) b} \hat y dx $$

Where the original I is replaced with I/b, for integrating over K should result in I, and it does (and it's what we were given). The biggest difference is (a-x), since the radius is based on the difference, not the variable directly. 

$$ \vec B = \frac{\mu_0 I}{2\pi b}\left[ -ln|x-a| \right]_{-b/2}^{b/2} \hat y  $$
$$ = \frac{\mu_0 I}{2\pi b}(-ln|b/2-a| + ln|-b/2-a|)\hat y $$

As a is always larger than b/2, the absolute values can be removed. 

$$ = \frac{\mu_0 I}{2\pi b}(ln(\frac{b/2+a}{a-b/2}))\hat y $$

Now, we solve for the y case. The integral is simlar. One might think we need to work in two dimensions of integral now, but this is not the case. Any y contributions to the field will be completely canceled by another wire--specifically, one diametrically opposite its position on the x-axis. Thus, all we need to worry about are the x components scaled based on the angle the wire makes with the point of interest. It'll scale specifically with the cosine of the angle, our old friend "adjacent over hypotenuse," which in this case is $\frac{a}{\sqrt{x^2+a^2}}$. The denominator here *also* happens to be the radius. So the integral becomes...

$$ \vec B = - \int_{-b/2}^{b/2} \frac{a}{\sqrt{x^2+a^2}}\frac{\mu_0 I}{2\pi \sqrt{x^2+a^2} b} \hat x dx $$

We'll notice that this rather quickly simplifies.

$$ \vec B = - \int_{-b/2}^{b/2} \frac{\mu_0 I a}{2\pi (x^2+a^2) b} \hat x dx $$

Which turns out to be a trig-related integral, which isn't all that surprising. 

$$ = - \frac{\mu_0 I a}{2\pi b}\left[ \frac{tan^{-1}(x/a)}{a} \right]_{-b/2}^{b/2} \hat x $$
$$ = - \frac{\mu_0 I}{\pi b} tan^{-1}(b/2a) \hat x$$
  
Which points purely in the -x direction as it should. Though we kind of assumed this at the start of the problem due to the symmetry argument, so this isn't as good of a check as it usually is. 

Now that we have our answers we need to do analysis on them. As a reminder, we have:

$$\vec B(0,a) = - \frac{\mu_0 I}{\pi b} tan^{-1}(b/2a) \hat x$$
$$\vec B(a,0) = \frac{\mu_0 I}{2\pi b}(ln(\frac{b/2+a}{a-b/2}))\hat y $$

Both of these expressions, to first order, vanish as a approaches infinity, which means we need to expand and find the first nonzero term. However, we should expand in terms of b/a not a directly like we were trying to do first.

$$\vec B(0,a) = - \frac{\mu_0 I}{\pi b} tan^{-1}(b/2a) \hat x$$
$$\vec B(a,0) = \frac{\mu_0 I}{2\pi b}(ln(\frac{b/a2+1}{1-b/2a}))\hat y $$

For small angles, atan(x) and tan(x) are said to equal x. b/a gets really small really quick, so this works for the first one.

$$\vec B(0,a) = - \frac{\mu_0 I}{2\pi a} \hat x$$

And we use the Taylor Expansion around b/a=0 on the natural log. This reduced to just... b/a, just like the above. Interesting.

$$ \vec B(a,0) = \frac{\mu_0 I}{2\pi a} \hat y$$

Which is to say that, at a distance, the plate looks just like a line current. Which makes sense, so long as a >> b. 

We're also asked the minimization question, that is, what value of b makes the value of the funciton lowest? Well, for the y version, this is accomplished by making b as large as possible. This makes sense becuase the larger we make it, the more the total current gets spread out, and the less strength it will have. 

As for the x version, making b larger makes it explode when b=a, but it is minimized when b=0. Notably it's still not zero here (note the discontinuity at x=0, it's not an infintie one), as there is still a line current at this point. 

So for a small field at (a,0) we want a single wire, but for a small field at (0,a) we want to spread out the charge as much as you possibly can along a very wide plate. Here's the two equations graphed with some specific values with the x-axis being values of b. 



The "U" is the (a,0) point, and the "hump" is the (0,a) point. Yes negative b values make no sense but it's easier to see the shape of the graphs this way. 

This uses some of the same principles as before. That said, instead of using the wire field, we use the loop field. The notes give this as:

$$ \vec B = \frac12 \mu_0 I \frac{R^2}{(z^2+R^2)^{3/2}} \hat z $$

Helpful thing to realize: the magnetic field points the same direction above and below the loop, so unlike with the electric field we don't have to worry about flipping the rings as we move around. Neat! 

So, since we're calculating for the center of this finite solenoid, let's let that center be the origin at z=0. This means the length L can be considered a range from -L/2 to L/2, which will be the bounds of our integral. Each individual "ring" (or loop) carries I, so we don't have to worry about chnging I. However, as we integrate from -L/2 to L/2, we do need to make sure we end up with total current at the end. The total current in the entire solenoid would be InL. To include this in the integral, all we do is replace I with In, which would effectively be the current density per unit length, which is what we're integrating over. 

The other consideration is how to manage z, since we're integrating over it. The best part is, since we're integrating from -L/2 to L/2, we don't need to, as the numbers we plug in already give us the exact contribution from one loop at that point. 

$$ \vec B = \int_{-L/2}^{L/2} \frac12 \mu_0 In \frac{R^2}{(z^2+R^2)^{3/2}} \hat z dz $$
$$ = \frac12 \mu_0 In R^2 \int_{-L/2}^{L/2} \frac{1}{(z^2+R^2)^{3/2}} \hat z dz $$
$$ = \frac12 \mu_0 In R^2 \left[ \frac{z}{R^2(z^2+R^2)^{1/2}} \right]_{-L/2}^{L/2} \hat z $$
$$ = \frac12 \mu_0 InL \frac{1}{((L/2)^2+R^2)^{1/2}} \hat z $$

Desirably, we have the total current, InL, which is a good sign. The current is definitely pointing in the right direction. 

There is a curious question as to what happens if we increase the length seemingly without bound? Well, in that case, we end up with L/2 dominating... Which means the fraction becomes 2/L, which cancels with the front terms leaving us only with $\mu_0 In$. Which is, in fact, the exact constant magnetic field that an infinite solenoid should have. Neat!
