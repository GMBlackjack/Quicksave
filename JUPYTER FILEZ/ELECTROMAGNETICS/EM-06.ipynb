# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

So, this is actually a very similar problem to the provided Box example problem in the notes. So we shall start from there. The base assumption is that $\varphi = XYZ$, where each portion is a funciton of one variable. The X and Y situations are exactly the same: functions bounded at 0 and L, forced to be zero (grounded). Z, however, is different, as it does not go from a potential to grounded, it goes from a potential to an endless limit. However, we can assume it exponentially decays since we have to go out to infinity. Thus, our base solutions are:

$$ X = A_n sin\frac{n\pi x}{L} $$
$$ Y = B_m sin\frac{m\pi y}{L} $$
$$ Z = C_{nm} e^{-\gamma z} $$

Where $\gamma^2 = \frac{(n^2+m^2)\pi^2}{L^2} $

So, combined all together, we have a solution of:

$$ \varphi = \sum_{n,m=1}^\infty C_{nm}sin\frac{n\pi x}{L} sin\frac{m\pi y}{L}e^{-\gamma z} $$

So to fully solve the posed quesiton we need to find the constant C, which is where most of the problem actually lies. Fortunately, we do have a boundary condition at z=0. However, it's piecewise, so for now we shall just treat it generically as f(x,y).

$$ f(x,y) = \sum_{n,m=1}^\infty C_{nm}sin\frac{n\pi x}{L} sin\frac{m\pi y}{L}$$
 

We can abuse orthogonality in order to solve for the constant. Just multiply by the same sine functions, except let n=n' and m=m'. The integral is known, from the notes, to be L/2 for each, so the total result of the orthogonal integrals is $L^2/4$ and gets rid of the sum on the right hand side. The left hand side has to keep the integrals, like so:

$$ \int_0^L \int_0^L f(x,y) sin\frac{n'\pi x}{L} sin\frac{m'\pi y}{L}  dx dy = C_{n'm'}\frac{L^2}{4}$$
$$ \Rightarrow \frac{4}{L^2} \int_0^L \int_0^L f(x,y) sin\frac{n\pi x}{L} sin\frac{m\pi y}{L}  dx dy = C_{nm}$$
We removed the primes since there is no longer an ambiguity. 

Now, we actually evaluate, by noting that f(x,y) is piecewise. It acutally only depends on y, biing -V from 0 to L/2 and V from L/2 to L. This means we can integrate over x first without splitting it up. 

$$ C_{mn} =\frac{4}{L^2} \int_0^L f(y) sin\frac{m\pi y}{L} \int_0^L  sin\frac{n\pi x}{L} dx dy$$
$$  =\frac{4}{L^2} \int_0^L f(y) sin\frac{m\pi y}{L} \left[ -\frac{L}{n\pi} cos\frac{n\pi x}{L} \right]_0^L dy$$
$$  =\frac{4}{L^2} \int_0^L f(y) sin\frac{m\pi y}{L} \frac{L}{n\pi} (1-cos(n\pi)) dy$$
$$  =\frac{4}{Ln\pi} (1-cos(n\pi)) \int_0^L f(y) sin\frac{m\pi y}{L} dy$$

Now we perform the splitting of the integral. 

$$  =\frac{4}{Ln\pi} (1-cos(n\pi)) \left( \int_0^{L/2} -V sin\frac{m\pi y}{L} dy + \int_{L/2}^L V sin\frac{m\pi y}{L} dy \right)$$
$$  =\frac{4V}{Ln\pi} (1-cos(n\pi)) \left( \left[ \frac{L}{m\pi}cos\frac{m\pi y}{L} \right]_0^{L/2} + \left[ -\frac{L}{m\pi}cos\frac{m\pi y}{L} \right]_{L/2}^{L} \right)$$
$$  =\frac{4V}{Ln\pi} (1-cos(n\pi)) \left( \left[ \frac{L}{m\pi}cos\frac{m\pi}{2} - \frac{L}{m\pi} \right] + \left[ \frac{L}{m\pi}cos\frac{m\pi}{2} -\frac{L}{m\pi}cos(m\pi) \right]\right)$$

$$  =\frac{4V}{nm\pi^2} (1-cos(n\pi)) \left( 2cos\frac{m\pi}{2} - cos(m\pi) - 1 \right)$$

This is the value for $C_{nm}$ so our full solution is:

$$ \varphi = \sum_{n,m=1}^\infty \frac{4V}{nm\pi^2} (1-cos(n\pi)) \left( 2cos\frac{m\pi}{2} - cos(m\pi) - 1 \right) sin\frac{n\pi x}{L} sin\frac{m\pi y}{L}e^{-\gamma z} $$

Now the behavior of this constant is a bit itneresting. The 1-cos portion has two values: 2 and 0, alternating on odd and even. Thus, only odd values of n exist.

The values of m are a bit different, since the cycle exists on a 4-cycle rather than a 2. That cycle is, starting from m=0...

2 - 1 - 1 = 0 

0 + 1 - 1 = 0

-2  -1  -1 = -4

0 + 1  - 1 = 0

So the first m that exists is m=2, then m=6, 10, etc. 

Note that the value of the coefficeint here, -4, would be +4 if we had defined the +V and -V potential to be in reverse, but the problem did not specify which was closer to x=0 (and would be physically the same, just in a different coordiante system, anyway). 



Now, we are asked to identify the term that dominates the furthest from the plate. That will depend on $\gamma$, which in turn depends on n and m. The most dominating term would have the smallest $\gamma$, and the smallest one that exists would depend on n=1 and m=2. Thus the dominating term is...

$$ \varphi_{dominate} = \frac{-16V}{\pi^2} sin\frac{\pi x}{L} sin\frac{2\pi y}{L}e^{-\sqrt{5}\pi z/L} $$

So, the last thing we need to find out is the asymptotic behavior, that is, the behavior when z gets really really big. And, as it turns out, all this is askign for is the specific form of the dominating term. Which we have listed above. So we're already done. It oscillates once in x, as is normal for something between two grounded potentials, but it oscillates "twice" in y, which makes sense because of the divide when y=L/2 between the two kinds of potentials. It also decays rather nicely.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

A note: we want the ELECTRIC FIELD to be as constant as possible, not the potential, thus we want an electric field that varies linearly with z (the easiest way to do this anyway). We declare that the seeming trivial answer of $\theta = 0$ (making both the charged sides of the shell go away) is not considered, as it wouldn't really be the same system now would it? 

The general form for ANY (separable) spherical potential is

$$ \varphi = \sum_{lm} \left( A_{lm} r^l + B_l \frac{1}{r^{l+1}}\right) Y_{lm} $$

We can immediately simplify this since we are concerned with the interior potential, removing the B constant.

$$ \varphi = \sum_{lm} A_{lm} r^l Y_{lm} $$

Also, we have azimuthal symmetry, that is, no $\phi$ dependence, so m=0.

$$ \varphi = \sum_{l} A_{l} r^l Y_{l0} $$

The boundary conditions are given ar r=R.

$$ f(\theta) = \sum_{l} A_{l} R^l Y_{l0} $$

The orthonormality has already been used in the notes to find the expression for the constant.

$$ A_l = \frac{1}{R^l} \int f(\theta) Y^*_{l0} d\Omega $$

Since m=0 we have no imaginary part to speak of so the conjugation goes away.

$$ A_l = \frac{1}{R^l} \int f(\theta) Y_{l0} d\Omega $$

Note that we can write $d\Omega = dcos\theta d\phi$. Integrating the $\phi$ produces a term of $2\pi$.

The problem is we can't really abuse orthogonality here again, sicne the potential is piecewise and isn't quite going to line up with any of the spherical harmonics. One thing we can say, however, is that the middle potential (the part that's grounded) will integrate to zero, so we can fully define the cosntatn as...

$$ A_l = \frac{2\pi}{R^l} \left( \int_1^{cos\theta_0} V Y_{l0} dcos\theta + \int_{-cos\theta_0}^{-1} -V Y_{l0} dcos\theta\right)  $$
$$ = \frac{2\pi V}{R^l} \left( \int_1^{cos\theta_0} Y_{l0} dcos\theta - \int_{-cos\theta_0}^{-1} Y_{l0} dcos\theta\right)  $$

Now, the m=0 spherical harmonics are given by:

$$ Y_{l0} = \sqrt{\frac{2l+1}{4\pi}}P_l(cos\theta) $$

where

$$ P_l = \frac{1}{2^l l!} \frac{d^l}{dx^l}(x^2-1)^l $$

Now, one thing we do know is that, unlike other coordinate systems, the spherical harmonics split up neatly into successive approximation terms. So if we're asked for the MOST constant, we simply look for what makes it constant at the first term to actually exist. So, let's study l=0 first.

$$ A_0  = \frac{2\pi V}{R^0} \left( \int_1^{cos\theta_0} \frac{1}{\sqrt{4\pi}} dcos\theta - \int_{-cos\theta_0}^{-1} \frac{1}{\sqrt{4\pi}} dcos\theta\right)  $$

Which resolves to zero, so this was a non-starter. Next...

$$ A_1  = \frac{2\pi V}{R^1} \sqrt{\frac{3}{4\pi}} \left( \int_1^{cos\theta_0} cos\theta dcos\theta - \int_{-cos\theta_0}^{-1} cos\theta dcos\theta\right)  $$

This resolves somewaht surprisingly to

$$ A_1 = - \frac{2\pi V}{R} \sqrt{\frac{3}{4\pi}} (sin^2\theta_0) $$

This is our first term, and thus it is our dominating term (it is specifically the dipole term). The dipole term in particular appears as

$$ \varphi_1 = A_1' r cos\theta $$

Where the prime denotes additional constants thrown in when the harmonic was added back. $rcos\theta = z$ which means that, if the system were a pure dipole, the electric field would be exactly constant. This reveals our ultimate goal: find the angle that makes this as pure of a dipole as possible. (Our guess is that $\theta = \pi/2$ is the correct one, that is it gets closer to a dipole the smaller we make the grounded segment, but we shall see). 

This does mean that we have to look at the larger harmonics and see how to minimize them. Fortunately for us, we know a few things about the polynomials $P_l$ that make up the harmonics. First of all, any constant term in these polynomials will go to zero, as we saw with $A_0$, and any linear term will go like the square sine. So what of higher order terms? Well, that all depends on the evaluateions of...

$$ \int_1^{cos\theta_0} cos^l\theta dcos\theta - \int_{-cos\theta_0}^{-1} cos^l\theta dcos\theta $$

For each P-polynomial only consists of various powers of cosine in our case, and all lower values of cosine will simply contribute in the same way as previous parts (albeit smaller due to the scaling by 1/l!). Regardless, the evaluation is...

$$ \frac{1}{l+1}(2cos^{l+1}\theta_0 - 1 + (-1)^{l+1}) $$

Note the behavior of this: for odd l, the 1s combine, but for even l, the 1s cancel.

$$ \frac{1}{l+1}(2cos^{l+1}\theta_0 - 2) $$
$$ \frac{1}{l+1}(2cos^{l+1}\theta_0) $$

Now we can easily remove the even or the odd solutions by setting $\theta$ a particular way. However, l=1 is an odd solution, and it's one we want to keep! So we choose to kill the even solutions (which includes the second most dominant term AFTER the dipole term, conveniently). This happens when $\theta_0 = \pi/2$. Which makes sense, as that is when our $A_1$ hits its maximum. 

This means with $\theta_0=\pi/2$ we have no normal term, a dipole term, no quadrupole term, and then an octupole term. IF we were to adjust the angle at all, we would end up with some quadrupole terms, which would counteract any lowering we got in the higher order approximations. And plus, we  rarely extend beyond quadrupole anyway, so if we consider quadrupole the end of it all, we have *nothing* but a dipole. That said, this is only when $\theta_0 = \pi/2$

When it equals $\pi/2$ we get a familiar term from the notes.

$$ \varphi = \frac{3}{2} V \frac{r cos\theta}{R} = \frac{3}{2} V \frac{z}{R} $$

And if we take the negative gradient...

$$ \vec E = - \frac{3}{2} V \frac{1}{R}\hat z $$

Which is constant. (Again, only to the quadrupole approximation). This was what we predicted, where the grounded center segment vanishes. Strictly speaking, this might not be the exact same system, but at least it isn't the trivial case wehre everything's grounded so there's functionally no charge.

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$

Okay so for this problem, no $\phi$ dependence, so we only have to worry about R and Z. That said the lack of G does tell us that $\alpha=0$, which will help us in determining some functions later.

So, first of all, the ends of the can are held at potential V. We happen to know that an *open* can's sides experience exponential decay, so naturally two different ends must experience exponential decay that feed into each other. To demonstrate what that is mathematically, we set the can with its center at the origin, making its length range from -L/2 to L/2. This means the potential forms of Z add together to take the form:

$$ Z = Ae^{-k(L/2 + z)} + Ae^{-k(L/2 - z)} $$
$$ = Ae^{-kL/2}e^{-kz} + Ae^{-kL/2}e^{kz} $$
$$ = Ae^{-kL/2} cosh(kz) $$

Which is to say Z is a cosh function. Which makes absolutely perfect sense, as there is no reason for the potential to drop to zero in the middle, and no reason for it to oscillate as well since the walls are grounded, not at *zero*. This means that $k^2>0$, for the record. 

For the record, the constant A is the same on both terms due to symmetry, the influence from both sides has to be completely equal and symmetric. 

Now we deal with R. Since $k^2$ is positive, R must take the Bessel Functions of the first and second kind. We need a finite potential at the center of the radius, though, so $N_\alpha$ is not an option, leaving us with $J_\alpha$ or $J_0$ since we already established $\alpha=0$

Thus, R is some constant times $J_0$, specifically $J_0(k\rho)$. 

Combining it all together, we have:

$$ \varphi = \sum_{n}^\infty C_n J_0(k_n \rho)e^{-k_nL/2} cosh(k_n z) $$

Now, we actually have *two* restrictions the constants $C_n$ have to meet. at z=$\pm L/2$, the potential has to go to V. However, at $\rho = R_0$ (the radius) the potential has to go to zero. Which ones do we use? Well, that would be the one where $J_0$ goes to zero, which have to be looked up in a table somewhere, naturally. Fortunately, we do not need these values, we can just exploit Bessel orthogonality and get a result. Set z=L/2 to get the boundary condition, and then integrate over radius and exploit the orthogonality. 

$$ V = \sum_n^\infty C_n J_0(k_n\rho) e^{-k_nL/2} cosh(k_n L /2)$$

$$ \Rightarrow \int_0^R V J_0(k_n' \rho) \rho d\rho= \int_0^R \sum_n^\infty C_n J_0(k_n\rho) e^{-k_nL/2} cosh(k_n L /2) J_0(k_n' \rho) \rho d\rho$$

Which causes the usual "only n=n' exists" result...

$$ \Rightarrow \int_0^R V J_0(k_n' \rho) \rho d\rho= C_n e^{-k_nL/2} cosh(k_n L /2) \frac{R^2}{2}(J_1(k_n R))^2 $$
$$ \Rightarrow \frac{1}{e^{-k_nL/2} cosh(k_n L /2) \frac{R^2}{2}(J_1(k_n R))^2}\int_0^R V J_0(k_n'\rho) \rho d\rho= C_n $$

Which is definitely ugly but that's probably why we don't have to evaluate it. 

So, the two ends of the can are grounded. cosh, sinh, and the real exponentials aren't going to cut it here, so we better have trig functions sin and cos. In order to make it easy on us we will set one of the can's ends at the origin, so the can ranges from 0 to L. This is so we only have sine functions (aka, ones that need to be zero at z=0). So Z is going to be some function of sines. And as in part a) the $\phi$ dependence is nonexistent. 

What about the radial portion, R? Well, since $k^2<0$ due to the forced sinusoidal nature of Z, then we know we have the Modified Bessel Functions $I_0, K_0$. Since K explodes at zero, we will have to resort to I. Thus, our whole potential is something of the form:

$$ \varphi = \sum_n^\infty C_n sin(k_nz) I_0(k_n\rho) $$

Now, curiously, we must recognize that it is the trig function, not I, that is limiting our possible solutions. I has no restrictions on it, it can go all over the place and still match everything, but only certain values of n will satisfy the sine function being zero at both ends. The thing is we already know what those values are from doign so many trig problems, no need to even work them out or anything, just fill them in.

$$ \varphi = \sum_n^\infty C_n sin(\frac{n\pi}{L}z) I_0(\frac{n\pi}{L}\rho) $$

Our boundary is V at $\rho=R$. 

$$ V = \sum_n^\infty C_n sin(\frac{n\pi}{L}z) I_0(\frac{n\pi}{L}R) $$

Now as far as integration with respect to z is concerned, I is a constant, so it doesn't change at all the normalization we use for the normal trig situations--which is to say the Fourier Sine series. Thus...

$$ C_n = \frac{2V}{L} I_0(\frac{n\pi R}{L}) \int^{L}_{0} sin(\frac{n\pi z}{L}) dz $$

$$ C_n = \frac{2V}{n \pi} I_0(\frac{n\pi R}{L}) (cos(n\pi) - 1) $$

Which is our answer, and is zero on even n. There we go!
