# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

So to discuss this, we should examine what is known about electric and magnetic dipole behavior in isolation.

$$ \vec B = -\frac{\mu_0}{4\pi c} \frac{\hat r \times \vec p ''}{r} $$
$$ \vec E = -\frac{\mu_0}{4\pi} \frac{\hat r (\hat r \cdot \vec p '') - \vec p''}{r} $$
$$ \frac{dP}{d\Omega} = \frac{\mu_0}{16\pi^2 c} |\hat r \times \vec p ''|^2 $$

$$ \vec E = \frac{\mu_0}{4\pi c} \frac{\hat r \times \vec m ''}{r} $$
$$ \vec B = -\frac{\mu_0}{4\pi c^2} \frac{\hat r (\hat r \cdot \vec m '') - \vec m''}{r} $$
$$ \frac{dP}{d\Omega} = \frac{\mu_0}{16\pi^2 c^3} |\hat r \times \vec m ''|^2 $$


Our goal is essentially to show that you can't just add the two power terms together. You can, as always, add electric and magnetic fields, though. One thing we can do is consider our unique situation, and make direct substitutions into the six equations above. The results should not be that hard to determine, as all taking the double derivative does is add a minus sign and a $\omega^2$ term.

$$ \vec B = \frac{\mu_0\omega^2}{4\pi c} cos(\omega t) \frac{\hat r \times \vec p_0}{r} $$
$$ \vec E = \frac{\mu_0\omega^2}{4\pi} cos(\omega t) \frac{\hat r (\hat r \cdot \vec p_0) - \vec p_0}{r} $$
$$ \frac{dP}{d\Omega} = \frac{\mu_0\omega^4}{16\pi^2 c} cos^2(\omega t) |\hat r \times \vec p_0|^2 $$

$$ \vec E = -\frac{\mu_0 \omega^2}{4\pi c} cos(\omega t) \frac{\hat r \times \vec m_0}{r} $$
$$ \vec B = \frac{\mu_0 \omega^2}{4\pi c^2} cos(\omega t) \frac{\hat r (\hat r \cdot \vec m_0) - \vec m_0}{r} $$
$$ \frac{dP}{d\Omega} = \frac{\mu_0\omega^4}{16\pi^2 c^3}cos^2(\omega t) |\hat r \times \vec m_0|^2 $$


So we add everythihng together, noting that what we sum as the power is explicitly *not* what we expect to actually get.

$$ \vec E = \frac{\mu_0 \omega^2}{4\pi r} cos(\omega t) \left[ \hat r (\hat r \cdot \vec p_0) - \vec p_0 - \frac1c \hat r \times \vec m_0\right]  $$

$$ \vec B = \frac{\mu_0 \omega^2}{4\pi c r} cos(\omega t) \left[ \frac1c [\hat r (\hat r \cdot \vec m_0) - \vec m_0] + \hat r \times \vec p_0\right]  $$

$$ \frac{dP}{d\Omega} \neq \frac{\mu_0\omega^4}{16\pi^2 c}cos^2(\omega t) \left[ \frac{1}{c^2}|\hat r \times \vec m_0|^2 + |\hat r \times \vec p_0|^2 \right] $$

Rather obbviously the P we added there does not have cross terms that we *will* have as a result of actually taking the Poynting vector, already it's obvious that powers don't add. However, we are asked more details than that, specifically asking for cases where we *do* get the powers to add. (Likely just by coincidence). 

Anyway, taking the full Poynting vector is basically impsosible since we have no idea where the various moments are pointing. So, instead, we shall rely on the trick that allows us not to take that cross product.

$$ \vec S = \frac{c}{\mu_0} B^2\hat r $$

We could have also done this with E, but B is the field of choice in the notes for representing dP, so why change it?

$$ \vec S = \frac{\mu_0 \omega^4}{16\pi^2 c} cos^2(\omega t) \frac{1}{r^2} \left[ \frac1c [\hat r (\hat r \cdot \vec m_0) - \vec m_0] + \hat r \times \vec p_0\right]^2  $$

Now this is simlar to our fake "added power" above, but we have a stray $1/r^2$ term. However, keep in mind that what we're actually going for is

$$ \frac{dP}{d\Omega} = r^2\vec S \cdot \hat r $$

Which immediately gets rid of our annoying radius term.

$$ \frac{dP}{d\Omega} = \frac{\mu_0 \omega^4}{16\pi^2 c} cos^2(\omega t) \left( \left[ \frac1c [\hat r (\hat r \cdot \vec m_0) - \vec m_0] + \hat r \times \vec p_0\right]^2  \right) \cdot \hat r $$
$$ \Rightarrow \frac{dP}{d\Omega} = \frac{\mu_0 \omega^4}{16\pi^2 c} cos^2(\omega t) \left( \left| \frac1c [\hat r (\hat r \cdot \vec m_0) - \vec m_0] + \hat r \times \vec p_0\right|^2  \right) $$

Now we just need to see what the term in parenthesis actually *is*.

$$ \Rightarrow \frac{dP}{d\Omega} = \frac{\mu_0 \omega^4}{16\pi^2 c} cos^2(\omega t) \left( \frac{1}{c^2} |[\hat r (\hat r \cdot \vec m_0) - \vec m_0]|^2 +\frac2c |([\hat r (\hat r \cdot \vec m_0) - \vec m_0]) \cdot (\hat r \times \vec p_0)| + |\hat r \times \vec p_0|^2  \right) $$

Behold the very obvious and obnoxious cross term, proof that adding does not occur. By the way, we can see that the cross term will go away when the dot product that defines it goes to zero. That is, **Interference does not occur when the field contributions are perpendicular to each other.** That is, B from magnetic dipole is always 90 degrees out of phase from the B from the electric dipole. This essentially means that the two types of radiation the dipoles are producing are being emitted in entirely different planes. 

We can actually take the time average now, all it does is introduce a factor of $1/2$.

$$ \Rightarrow \langle \frac{dP}{d\Omega}\rangle = \frac{\mu_0 \omega^4}{32\pi^2 c} \left( \frac{1}{c^2} |[\hat r (\hat r \cdot \vec m_0) - \vec m_0]|^2 +\frac2c |([\hat r (\hat r \cdot \vec m_0) - \vec m_0]) \cdot (\hat r \times \vec p_0)| + |\hat r \times \vec p_0|^2  \right) $$

At first, it seems like the first term will just up and simplify to a cross product. Let's show this.
$$ |[\hat r (\hat r \cdot \vec m_0) - \vec m_0]|^2 $$
$$ = |[\hat r (\hat r \cdot \vec m_0) - \vec m_0(\hat r \cdot \hat r)]|^2 $$
$$ = |\hat r \times (\hat r \times \vec m_0)|^2 $$
Note that crossing by the same unit vector twice does nto actually change the magnitude as the first cross product makes it orthogonal. So...
$$ = |(\hat r \times \vec m_0)|^2 $$

And thus...

$$ \Rightarrow \langle \frac{dP}{d\Omega}\rangle = \frac{\mu_0 \omega^4}{32\pi^2 c} \left( \frac{1}{c^2} |\hat r \times \vec m_0|^2 +\frac2c |([\hat r (\hat r \cdot \vec m_0) - \vec m_0]) \cdot (\hat r \times \vec p_0)| + |\hat r \times \vec p_0|^2  \right) $$

To demonstrate this, we need to show that the integral over solid angle of the cross term goes to zero.

$$ \int \frac2c |([\hat r (\hat r \cdot \vec m_0) - \vec m_0]) \cdot (\hat r \times \vec p_0)| sin\theta d\theta d\phi $$

Since we have so many unknown vectors we will naturally seek an argument that this has to go to zero rather than a manual evaluation. AS such we can ignore the leading constant.

$$ \int |([\hat r (\hat r \cdot \vec m_0) - \vec m_0]) \cdot (\hat r \times \vec p_0)| sin\theta d\theta d\phi $$

Let's expand out the dot product.

$$ = \int |(\hat r (\hat r \cdot \vec m_0)) \cdot (\hat r \times \vec p_0) - (\vec m_0) \cdot (\hat r \times \vec p_0)| sin\theta d\theta d\phi $$

That first dot product goes to zero, as we are dotting a radial vector into the cross product of the radial vector. Or, more accuratley, anything of the form $a\cdot (a \times b)$ goes to zero. 

$$ = -\int | (\vec m_0) \cdot (\hat r \times \vec p_0)| sin\theta d\theta d\phi $$

Now, note that the vectors for m and p are constant, but we're integrating over *all* possible r directions. This means that for every possible result, we can also find an r direction that has the exact *opposite*. As we add everything together, those terms will cancel, leaving us with zero. Thus, when we integrate over the entire solid angle, the cross term goes away.
