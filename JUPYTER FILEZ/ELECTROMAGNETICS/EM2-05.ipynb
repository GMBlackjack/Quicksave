# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

The Drude model for conductors is

$$ m\frac{d\vec v}{dt} = q\vec E - \frac{m}{\tau}\vec v $$

Now we are told we have a tamping term here, so that remains the same. That electric field term comes from the EM wave itself, so it also remains. However, we have a magnetic term instead of an electric term. We know that the magnetic contribution to the force is a cross product, v cross B, so we re-write this as...

$$ m\frac{d\vec v}{dt} = q\vec E + q(\vec v \times \vec B) - \frac{m}{\tau}\vec v $$

We have some specific values in our case. q = -e accordint to the problem, and our magnetic field is purely in the z direction. We may think that since the waves are propogating parallel to the magnetic field that the cross product would go to zero, but actually the propogation direction is *perpendicular* to the actual direction of the EM field, so we'll need to be careful. 

$$ m\frac{d\vec v}{dt} = -e(\vec E + \vec v \times B\hat z) - \frac{m}{\tau}\vec v $$

We will assume the fluctuations induced by the EM wave's magnetic field are neglegible compared to the overall applied magnetic field. 

Since the E field comes from the EM wave, we know it has to be time harmonic. In the original Drude model, this meant the velocities had to be harmonic. However, is this the case here, for there is an extra B term. 

The answer is that, yes, it does, For when we carry out the cross product (even not knowing the direction of v for sure yet) we end up with...

$$ m\frac{d\vec v}{dt} = -e(\vec E + vB (\hat v\times \hat z)) - \frac{m}{\tau}\vec v $$

Which means that if E is time harmonic, every other term with it better also be time harmonic so everything can canel. Since only v can be time harmonic among all the other variables (as B is constant) we must have a time harmonic v or the equaiton simply cannot hold. 

So, we declare the time harmonic equaiotns to be $\vec v = \vec v_0 e^{-i\omega t}$ and similarly for E. This means we can directly take the derivative...

$$- i m \omega \vec v = -e(\vec E + \vec v \times B\hat z) - \frac{m}{\tau}\vec v $$

The problem is that we can't just solve for v as a whole, since it's engaged in a cross product with the magnetic field. However, we do know B is only in the z direction, so we can write out the cross product as...

$$\vec v \times B \hat z = v_y B \hat x - v_x B \hat y$$

So the cross product is purely in x and y. As the EM is traveling in the z direction, we know it is restricted to the xy planea s well. One might think this means v itself has no z component, but let's be careful, the E and B terms could cancel each other out, leaving just the v terms. But we can still argue this physically: E never poitns in z, and magnetic fields can only apply forces perpendicular to their direction. Thus, there is no way a particle could ever start moving in the z direction unless it already *started* in motion, and that's not the situation we're considering right now. (The plasma is just a medium for the EM wave to travel through in our case).

So, v is purely in x and y. Which means that the equaiton has to hold for all its x directional componetns and all its y directional components. 

$$ \Rightarrow - i m \omega v_x \hat x = -e(E_x \hat x + v_y B \hat x) - \frac{m}{\tau}v_x \hat x $$
$$ \Rightarrow - i m \omega v_y \hat y = -e(E_y \hat y - v_x B \hat y) - \frac{m}{\tau}v_y \hat y $$

What our the E components? Well, since we are dealing with circularly polarized light, we have a polarization of $\hat x \pm i\hat y $, so...

$$ \Rightarrow - i m \omega v_x \hat x = -e(E \hat x + v_y B \hat x) - \frac{m}{\tau}v_x \hat x $$
$$ \Rightarrow - i m \omega v_y \hat y = -e(\pm i E \hat y - v_x B \hat y) - \frac{m}{\tau}v_y \hat y $$

Since we're looking at components, we can remove the unit vectors. 

$$ \Rightarrow - i m \omega v_x  = -e(E  + v_y B ) - \frac{m}{\tau}v_x  $$
$$ \Rightarrow - i m \omega v_y  = -e(\pm i E  - v_x B ) - \frac{m}{\tau}v_y  $$

Now, we have two equations and two unknowns, and those unwknowns are the two components of v. We can now solve for them. 

$$ v_y = \frac{-im\omega v_x + \frac{m}{\tau} v_x + eE}{-eB} $$
$$ v_x = \frac{-im\omega v_y + \frac{m}{\tau} v_y \pm ieE}{eB} $$
 
Annoyingly none of this simplifies so we're gonna have some complicated nonsense for a while here. 

Substitute...

$$ v_y = \frac{-im\omega \frac{-im\omega v_y + \frac{m}{\tau} v_y \pm ieE}{eB} + \frac{m}{\tau} \frac{-im\omega v_y + \frac{m}{\tau} v_y \pm ieE}{eB} + eE}{-eB} $$

Split this up into terms so we can actually see what we're looking at.

$$ \Rightarrow v_y = \frac{1}{-e^2B^2} \left[ -im\omega (-im\omega v_y + \frac{m}{\tau} v_y \pm ieE) + \frac{m}{\tau} (-im\omega v_y + \frac{m}{\tau} v_y \pm ieE) + e^2EB \right] $$ 
$$ \Rightarrow \frac{1}{-e^2B^2}\left[ \mp m\omega eE \mp \frac{m}{\tau} ieE - e^2EB \right] = \frac{1}{-e^2B^2} \left[ -im\omega (-im\omega v_y + \frac{m}{\tau} v_y) + \frac{m}{\tau} (-im\omega v_y + \frac{m}{\tau} v_y)\right] - v_y $$
$$ \Rightarrow \frac{\frac{1}{-e^2B^2}\left[ \mp m\omega eE \mp \frac{m}{\tau} ieE - e^2EB \right]}{\frac{1}{-e^2B^2} \left[ -im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau})\right] - 1} = v_y $$

Let's see what we can simplify...

$$ \Rightarrow \frac{ \mp m\omega eE \mp \frac{m}{\tau} ieE - e^2EB }{ -im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2} = v_y $$

The process for $v_x$ is very similar. 

$$ v_x = \frac{-im\omega \frac{-im\omega v_x + \frac{m}{\tau} v_x + eE}{-eB} + \frac{m}{\tau} \frac{-im\omega v_x + \frac{m}{\tau} v_x + eE}{-eB} \pm ieE}{eB} $$

$$ \Rightarrow v_x = \frac{1}{-e^2B^2} \left[ -im\omega (-im\omega v_x + \frac{m}{\tau} v_x + eE) + \frac{m}{\tau} (-im\omega v_x + \frac{m}{\tau} v_x + eE) \mp ie^2EB \right] $$ 

$$ \Rightarrow \frac{\frac{1}{-e^2B^2}\left[ im\omega eE - \frac{m}{\tau} eE \pm ie^2EB \right]}{\frac{1}{-e^2B^2} \left[ -im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau})\right] - 1} = v_x $$
$$ \Rightarrow \frac{ im\omega eE - \frac{m}{\tau} eE \pm ie^2EB }{-im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2} = v_x $$

Since $\vec j = nq\vec v = -ne\vec v$, we can now find the current components.

$$ -n \frac{ \mp m\omega e^2 \mp \frac{m}{\tau} ie^2 - e^3B }{ -im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2}E = j_y $$
$$ -n \frac{ im\omega e^2 - \frac{m}{\tau} e^2 \pm ie^3B }{-im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2}E = j_x $$

Note that even though we are still in componetns, we have the right form of "some funciton of $\omega$ multiplied by E." The presence of B though makes it rather difficult to see if there's anything else here, even taking the limit as $\omega -> 0$ doesn't get rid of the B term. 

We can still get a term out of this of the DC conductivity, $\sigma_0 = ne^2\tau/m$. 

$$ -\frac{ \mp mn\omega e^2 \mp \frac{m^2}{\tau^2} i\sigma_0 - ne^3B }{ -im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2}E = j_y $$
$$ -\frac{ imn\omega e^2 - \frac{m^2}{\tau^2} \sigma_0 \pm ine^3B }{-im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2}E = j_x $$

Now, $\vec j = \sigma \vec E$. Which means we need to extract back out our $E_x$ and $E_y$, which were just $E$ and $\pm iE$, respectively. 

$$ -\frac{ imn\omega e^2 - \frac{m^2}{\tau^2} \sigma_0 \pm i ne^3B }{ -im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2}(\pm i E) = j_y $$
$$ -\frac{ imn\omega e^2 - \frac{m^2}{\tau^2} \sigma_0 \pm ine^3B }{-im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2}E = j_x $$

The two terms are constant, as they should be, which means...

$$ \sigma = -\frac{ imn\omega e^2 - \frac{m^2}{\tau^2} \sigma_0 \pm ine^3B }{-im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2}$$

The fact that it was so easy to extract this suggests that maybe we didn't have to split up components so heavily, but since it just fell out it serves as a good sanity check that we didn't mess something up along the way. 

Regardless this also confirms that the conductivity is different for each polarized state as we still have a $\pm$ in this relation. Curiously, it only differs based on B, so if there was no B there would be no difference at all. 

Anyway this makes the final answer...

$$ \epsilon = \epsilon_0 - \frac{i}{\omega}\frac{ imn\omega e^2 - \frac{m^2}{\tau^2} \sigma_0 \pm ine^3B }{-im\omega (-im\omega + \frac{m}{\tau}) + \frac{m}{\tau} (-im\omega + \frac{m}{\tau}) + e^2B^2} $$

$$ \epsilon = \epsilon_0 - \frac{ -mn\omega e^2 - \frac{m^2}{\tau^2} i\sigma_0 \mp ne^3B }{-im\omega^2 (-im\omega + \frac{m}{\tau}) + \frac{m\omega}{\tau} (-im\omega + \frac{m}{\tau}) + e^2\omega B^2} $$

Which is a mess that's somewhat hard to get anything out of, admittedly. Notably this is different for both polarizations of light, but if B=0 that goes away, so at least that check can be performed. Trying to reduce this in $\omega$ led to something purely imaginary, but that wasn't very illuminating, so we end our checks here and move on to the next problem.

# Problem 2 \[Back to [top](#toc)\]
$$\label{P2}$$

We note that the wave provided to us is in a form that implies $k = \omega/c$, which indicates that the phase velocity is the speed of light outside the material. Which. Is not surprising at all this is a plane wave in empty space. We will also note that this is a plane wave so it's already going to be present within the material and on the other side.

Now, we need to be careful, but let's make an assumption here: that the wave that comes out the other side is the same as the wave that goes in. In a realistic situation there would be dissipation but we aren't considering that, and since there is no reflection, energy cannot be lost as the wave travels through. Thus, the form of the waves on both sides of the material must be the same. However, that does not mean the *phase* is the same, so while region I has the wave we are given, region III (the other side) might have a phase shift.

$$ \vec E_{III} = \hat y E_0 e^{i\omega(z/c-t+\frac{\delta}{\omega})} $$

Where $\delta$ could be any real constant. It's the phase shift.

Now, what about region II? This simply adds the index of refraction.

$$ \vec E_{II} = \hat y E_0 e^{i\omega(n(\omega)z/c-t+\frac{\delta}{\omega})} $$

And to complete the set...

$$ \vec E_{I} = \hat y E_0 e^{i\omega(z/c-t)} $$

We can pull out the phase shifts from the frequency-dependent terms. 

$$ \vec E_{I} = \hat y E_0 e^{i\omega(z/c-t)} $$
$$ \vec E_{II} = \hat y E_0 e^{i\delta_1} e^{i\omega(n(\omega)z/c-t)} $$
$$ \vec E_{III} = \hat y E_0 e^{i\delta_2} e^{i\omega(z/c-t)} $$

Now the frequency-dependent term is always in the exponent, and there's nothing we can do to change that, since the way it's stated hasa k in terms of $\omega$, making it clear that k depends on it. So there does not appear to be anything else we can do. 

The polarization and amplitude should not depend on $\omega$. Note that in our expresion above the phase shift does not explicitly depend on $\omega$, but it very likely does as the actual value of the $\delta values will most assuredly be determined by boundary conditions which will be set both by the length of the medium a and the frequency. 

Regions I and III do not depend on the index of refraction, so we don't change them.

$$ \vec E_{I} = \hat y E_0 e^{i\omega(z/c-t)} $$
$$ \vec E_{III} = \hat y E_0 e^{i\delta_2} e^{i\omega(z/c-t)} $$

Region II, on the other hand,

$$ \vec E_{II} = \hat y E_0 e^{i\delta_1} e^{i(\left[ \omega_0 n(\omega_0) + (c/v_g)(\omega-\omega_0)\right]z/c-\omega t)} $$ 

We shall leave this as is for now sicne we aren't entirely sure what we're going to do with it yet. 

So our expression for region I in packet language is then...

$$ \vec E_I = \hat y \int \hat E(z,w) e^{-i\omega t}d\omega = \hat y \int E_0 e^{i\omega z/c}e^{-i\omega t} d\omega $$

Similarly for the others...

$$ \vec E_{II} = \hat y \int E_0 e^{i\delta_1} e^{i(\left[ \omega_0 n(\omega_0) + (c/v_g)(\omega-\omega_0)\right]z/c)}e^{-i\omega t} d\omega $$
$$ \vec E_{III} = \hat y \int E_0 e^{i\delta_2} e^{i\omega z/c}e^{-i\omega t} d\omega $$

Now that we've substituted everything, we write it in $e^{i\phi} f(x)$, where x is a function of z and t. 

For the exterior waves, there is no exponential part, everything within the integral is f(x) since $E_0$ is considered part of it for part d. Also, the phases $\delta$ likely depend on $\omega$ in some way so we can't pull them out.

However, can we pull anything out of the second one?

$$ \vec E_{II} = \hat y e^{i\omega_0 n(\omega_0) z/c} e^{-i \omega_0 z/v_g } \int E_0  e^{i\delta_1} e^{i \omega z/v_g } e^{-i\omega t} d\omega $$

Yes, in fact, we can. Normally we wouldn't pull out $v_g$, but for this problem we can assume it is a constant.

Technically we've already done the writing down of the field in the previous part.

$$ \vec E_I = \hat y \int E_0 e^{i\omega z/c}e^{-i\omega t} d\omega $$
$$ \vec E_{II} = \hat y e^{i\omega_0 n(\omega_0) z/c} e^{-i \omega_0 z/v_g } \int E_0 e^{i\delta_1} e^{i \omega z/v_g } e^{-i\omega t} d\omega $$
$$ \vec E_{III} = \hat y \int E_0 e^{i\delta_2} e^{i\omega z/c}e^{-i\omega t} d\omega $$

Now, we should be able to match boundary conditions to get the result. The condition we have is

$$ \hat n \times (\vec E_1 - \vec E_2) = 0 $$

This is a cross product. The normal vector in this case is purely in the z direction, and both E terms are purely in the y direction. Curiously, we don't actually have to evaluate this product as we know it has to go to zero, so the actual E-vectors have to subtract to zero since the directions don't line up with the normal vector. 

$$ \vec E_I = \vec E_{II} $$ 

Though this only applies at the boundary, that is, z=0. With z=0, the relation becomes...

$$ \hat y \int E_0 e^{-i\omega t} d\omega = \vec E_{II} = \hat y \int E_0 e^{i\delta_1} e^{-i\omega t} d\omega $$

Which is only true if $\delta_1 = 0$. This updates our understanding of $\vec E_{II}$

$$ \vec E_{II} = \hat y e^{i\omega_0 n(\omega_0) z/c} e^{-i \omega_0 z/v_g } \int E_0 e^{i \omega z/v_g } e^{-i\omega t} d\omega $$

We have another boundary condition at z=a, between II and III.

$$ \hat y e^{i\omega_0 n(\omega_0) a/c} e^{-i \omega_0 a/v_g } \int E_0  e^{i \omega a/v_g } e^{-i\omega t} d\omega = \hat y \int E_0 e^{i\delta_2} e^{i\omega a/c}e^{-i\omega t} d\omega $$

Which is only true if

$$ e^{i\omega_0 n(\omega_0) a/c} e^{-i \omega_0 a/v_g } e^{i \omega a/v_g } = e^{i\delta_2} e^{i\omega a/c} $$

$$ \Rightarrow i\omega_0 n(\omega_0) a/c - i\omega_0 a/v_g + i\omega a/v_g = i\delta_2 + i\omega a/c  $$
$$ \Rightarrow \omega_0 n(\omega_0) a/c - \omega_0 a/v_g + \omega a/v_g = \delta_2 + \omega a/c  $$
$$ \Rightarrow \omega_0 n(\omega_0)/c - \omega_0/v_g + \omega/v_g = \delta_2/a + \omega/c  $$
$$ \Rightarrow (\omega - \omega_0)/v_g = \delta_2/a + (\omega- \omega_0 n(\omega_0))/c  $$
$$ \Rightarrow a(\omega - \omega_0)/v_g - a(\omega- \omega_0 n(\omega_0))/c = \delta_2  $$

This will change the equation for region III.

$$ \vec E_{III} = \hat y \int E_0 e^{ia((\omega - \omega_0)/v_g - (\omega- \omega_0 n(\omega_0))/c )} e^{i\omega z/c}e^{-i\omega t} d\omega $$

Now that we've solved for everything, we should be able to show what we were asked for. Let's try to show that the direction of propogation for region II is negative.

$$ \vec E_{II} = \hat y e^{i\omega_0 n(\omega_0) z/c}  e^{-i \omega_0 z/v_g } \int E_0 e^{i \omega z/v_g } e^{-i\omega t} d\omega $$


This basically equates to saying that k is negative. Note that we can easily extract k from the exponential term still in the integral, $\omega/v_g$. The drift velocity is negative and light frequencies are positive, so this quantity is negative, thus we are moving backward. 

Once we have determined that the packet in region II is moving backward, simple logic and boundary condition matching demands that the packet in region III appear before the packet in region I arrives at its boundary.

At some time, the packet in region II must appear at z=a. Since the boundary condition must match, at this time a packet must also be generated in region III. There is NO packet at Region I's boundary at this time since region II's packet is on the other side. Packet II thent ravels back to z=0, and once it reaches it, the boundary conditions must match so region I now has a packet at the interface, while there is now no packet at the interface between II and III.

Thus, III's packet must be genreated and emitted before I's packet reaches the interface, all because packet II is moving backward.
