# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Let us consider two universes, call them 1 and 2. Each of them have the two spheres just sitting in space, and they have the same geometries. We note that the relations must be true. No matter what the situation is, however, we end up with the following sums:

$$Q_i = \sum_j C_{ij} \varphi_j$$
$$\varphi_i = \sum_j P_{ij} Q_j$$

Which can essentially be written in matrix form as $Q = C\varphi; \varphi = PQ$. This notably indicates that P and C are inverses of each other. Since we are asked to find both, finding P and then C seems like a good idea.

So let's consider the potentials. Both conducting spheres appear as spherical shells of charge, which is to say the potentials inside them are constant (the potentials we are trying to find) and their charge is all on the surface. We are told that the spheres are far enough away that the charge densities remain uniform on the spheres, so we can treat each sphere as a point charge. If we assume R refers to the distance *from center to cneter* then we don't even have to worry about the radiuses of the spheres--they're far enough away that that particular geometry has no effect. Each interior potential is the addition of the potential from one sphere on itself and the other sphere. Since we are far enough away that we assume constancy, we don't have to worry about integrating here. The radiuses of the spheres themselves only matter for the self-interaction.

$$ \varphi_A = \frac{Q_A}{4\pi\epsilon_0 R_A} + \frac{Q_B}{4\pi \epsilon_0 R} $$
$$ \varphi_B = \frac{Q_A}{4\pi\epsilon_0 R} + \frac{Q_B}{4\pi \epsilon_0 R_B} $$

Which makes our potential matrix...

$$ P_{ij} = \frac{1}{4\pi\epsilon_0}\begin{bmatrix} 
\frac{1}{R_A} & \frac 1R \\
\frac 1R & \frac{1}{R_B} \\
\end{bmatrix}$$

And we use Wolfram to find the inverse. 

$$ C_{ij} = 4\pi\epsilon_0 \frac{R}{R_AR_B - R^2} \begin{bmatrix} 
-R_AR & R_AR_B \\
R_AR_B & -R_BR \\
\end{bmatrix}$$

And we're done!

Well to figure this out we're going to need to find some matrices.

$$Q_i = \sum_j C_{ij} \varphi_j$$
$$\varphi_i = \sum_j P_{ij} Q_j$$

Now one of the good things is that we're told *exactly* what the potential on each individual part is. Let's call the inner sphere 1, the lower hemisphere 2, and the upper hemisphere 3.

$$ \varphi_j = (0,0,V) $$

Next, we would seek to find the potential matrix. The various potentials due to the charged sphere are somewhat easy, as it is a perfect sphere. It interacts with itself via $\frac{Q_1}{4\pi\epsilon_0 a}$ and on both of the hemispheres with $\frac{Q_1}{4\pi\epsilon_0 b}$. However, that's only 3 out of 9 potentials. On page 137 of the textbook we find that the quantities of the capacitance matrix (and thust the potential matrix) are wholly symmetric, so we actually have 5 of the 9 components. However, perhaps this is enough information to work from? 

Using the first row of the potential matrix, we can write:

$$ 0 = \frac{Q_1}{4\pi\epsilon_0 a} + \frac{Q_2}{4\pi\epsilon_0 b} + \frac{Q_3}{4\pi\epsilon_0 b} $$

Now while we don't know the capacitance matrix, we do have two zero potentials, so we can write:

$$ Q_1 = VC_{13} $$
$$ Q_2 = VC_{23} $$
$$ Q_3 = VC_{33} $$

However this does not immediately restrict our degrees of freedom any. Thus, we're going to examine a second universe situation to see what we can learn via reciprocity. For a general other universe '...

$$ \varphi_1Q_1' + \varphi_2Q_2' + \varphi_3Q_3' = \varphi_1'Q_1 + \varphi_2'Q_2 + \varphi_3'Q_3 $$

Now we have a lot of zero potentials so we can reduce it to

$$ VQ_3' = \varphi_1'Q_1 + \varphi_2'Q_2 + \varphi_3'Q_3 $$

We are allowed to specify Q or $\varphi$ for each component in the primed universe, but not both for any component. Let's create a system with both of the outer hemispheres charged, but the interior at zero. That situation becomes identical to one of two hollow sphere conductors.

$$ VQ_3' = \varphi_2'Q_2 + \varphi_3'Q_3 $$

Now in this we can find some helpful things because technically we've already solved the primed universe's setup, the two spherical shells. The two hemispheres are one object "23". Note that the potential would be constant along the sphere, but the charge would be split in "half" overall. 

$$\varphi_2' = \varphi_3' = \frac{2Q_3'}{4\pi \epsilon_0} \frac{b-a}{b^2}$$
$$2Q_3' = \varphi_3' 4\pi\epsilon_0 \frac{b^2}{b-a}$$

Now we can freely specify what we want these to be. Setting both potentials to 1 seems to be the most helpful. This means...

$$2Q_3' = 4\pi\epsilon_0 \frac{b^2}{b-a}$$

And then...

$$ V2\pi\epsilon_0 \frac{b^2}{b-a} = Q_2 + Q_3 $$

This is actually useful, since we have these two things added together in our top relation.

$$ 0 = \frac{Q_1}{4\pi\epsilon_0 a} + \frac{Q_2}{4\pi\epsilon_0 b} + \frac{Q_3}{4\pi\epsilon_0 b} $$
$$ \Rightarrow 0 = \frac{Q_1}{4\pi\epsilon_0 a} + \frac{1}{4\pi\epsilon_0 b}V2\pi\epsilon_0 \frac{b^2}{b-a} $$

$$ \Rightarrow 0 = \frac{Q_1}{4\pi\epsilon_0 a} + \frac{bV}{2(b-a)} $$
$$ \Rightarrow -\frac{bV}{2(b-a)} = \frac{Q_1}{4\pi\epsilon_0 a}  $$
$$ \Rightarrow -\frac{2bV\pi \epsilon_0 a}{(b-a)} = Q_1 $$

And that's one of the charges we need! Actually, reading the question again, we realiaze that that is actually ALL we need. 

$$ V2\pi\epsilon_0 \frac{b^2}{b-a} = Q_2 + Q_3 $$
$$ -\frac{2bV\pi \epsilon_0 a}{(b-a)} = Q_1 $$

# Problem 3 \[Back to [top](#toc)\]
$$\label{P3}$$


So, rather than trying to integrate over the sphere, we are going to use a trick. That trick is the knowledge that 

$$\frac{\vec{F}}{A} = \frac{\sigma^2}{2\epsilon_0} \hat n$$

The charge density is the total charge divided by the surface area. Or $\frac{Q}{4 \pi R^2}$ Since the sphere is split in half we assume we are calculating immediately after it was split, and so the charge density remains constant for our purposes. 

So we now have...

$$\frac{\vec{F}}{A} = \frac{Q^2}{32 \pi^2 R^4 \epsilon_0} \hat n$$

Now one might think we just multiply both sides by A, but the normal vector depends on A so we have to actually integrate. 

$$ \vec F = \frac{Q^2}{32 \pi^2 R^4 \epsilon_0} \int \hat n dV $$

Integration factor...

$$ \vec F = \frac{Q^2}{32 \pi^2 R^4 \epsilon_0} \int \hat n \rho^2 cos\theta d\rho d\theta d\phi$$

We are not integrating over the radius, it is held constant.

$$ \vec F = \frac{Q^2}{32 \pi^2 R^2 \epsilon_0} \int \hat n cos\theta d\theta d\phi$$

Now we do the somewhat odd step of converting the normal unit vector into cartesian form so we can identify the directions going on here. 

$$ \vec F = \frac{Q^2}{32 \pi^2 R^2 \epsilon_0} \int (sin\theta cos\phi \hat x + sin \theta sin\phi \hat y + cos\theta hat z) cos\theta d\theta d\phi$$

Now the angle $\phi$ is integrated from 0 to 2$\pi$, thus making any standard trig funcitons with it go to zero. This includes both the x and y direction, leaving only z. 

$$ \vec F = \frac{Q^2}{32 \pi^2 R^2 \epsilon_0} \int_0^\pi (cos\theta hat z) cos\theta d\theta $$

Which goes to...

$$ \vec F = \frac{Q^2}{32 \pi^2 R^2 \epsilon_0} \frac{\pi}{2} \hat z$$

Which naturally makes sense that the hemisphere is being pushed upward in the z direction. The other hemisphere would go the opposite direction, just add a minus sign. 

$$ \vec F_1 = \frac{Q^2}{64 \pi R^2 \epsilon_0} \hat z$$
$$ \vec F_2 = -\frac{Q^2}{64 \pi R^2 \epsilon_0} \hat z$$

Equal and opposite reactions. 

The goal is to find a point charge that will act on one of the hemispheres with precisely the opposite force the second hemisphere tries to push it with. This means the charge needs to be negative to assert attraction. Also we kind of just assume point Q' is fixed even though that is quite unrealistic. 

So we have a simple question that needs answering: what is the force of a point charge on a hemisphere? (or vice-versa, we can use newton's law either way if one is easier to calculate.) One of the assumptions implicitly made on the previous part was that the interior had constant potential and no field. That no longer applies here, though the force from the other hemisphere must be the same. 

The original force relation was:

$$\frac{\vec F}{A} = \sigma \vec E$$

Now the reason we didn't use this in the previous one was because it was a conductor situation, where the field at the surface was undefined. But in the case of a point charge, this isn't the case, we know exactly what the electric field on it is! $\frac{Q'}{4\pi \epsilon_0 R^2} \hat n$ The annoying thing here is that $\hat n$ is not constant, just as in part a), so we are going to have to integrate again. 

$$\vec F = \sigma \frac{Q'}{4\pi \epsilon_0 R^2}\int \hat n dA $$

We already know this integral and the value of $\sigma$. 

$$ \vec F = \frac{Q}{4 \pi R^2} \frac{Q'}{4\pi \epsilon_0 R^2} R^2 \frac{\pi}{2} \hat z = \frac{QQ'}{32 \pi R^2 \epsilon_0} \hat z $$

Now we want this to be equal and opposite to the force the shell feels from the other shell. This can rather easily be seen to be when Q' = -Q/2, for it then becomes...

$$\vec F = -\frac{Q^2}{64 \pi R^2 \epsilon_0} \hat z $$

And we flip the sign for the other hemisphere as it is in the negative z direction. And that's the entire problem!

As a sanity check, this makes some sense. In effect, the charges of each hemisphere are half the total charge Q, so the charge that would "balance" them would also be half Q. The charge of the point is not "split" by the two hemispheres, both of them feel the same force from the point whether or not the other hemisphere exists, so only a half charge would need to balance it. 

Granted it's a little convenient that it's EXACTLY half a charge, but this is a spherical situation, so not all that surprising that it's neat. 

There is no change. At the instant of being cut, an insulator and a conductor act identically. Now over *TIME* there will be differences, but at the initial moment the charge distribution is identical and thus there's not anything to be concerned with.
