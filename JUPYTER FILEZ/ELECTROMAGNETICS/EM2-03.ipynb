# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

Two waves of equal amplitude traveling in opposite directions are well known to form a standing wave when added together. But presumably the form asked for is the one that is one sine wave with a time-dependent amplitude. Well, we can certainly do that with some trig rules...

$$ E_0 (sin(kz - \omega t) + sin(kz + \omega t)) \hat y$$
$$ = E_0 (sin(kz)cos(\omega t) - cos(kz)sin(\omega t) + sin(kz)cos(\omega t) + cos(kz)sin(\omega t)) \hat y$$
$$ = 2 E_0 sin(kz)cos(\omega t) \hat y $$

Which we submit as proper standing wave form.



We can get the magnetic field based on the derived rules for EM waves in empty space. 

First of all, $B_0 = \frac{k}{\omega} E_0$. This gives us our B-magnitude (save the factor of 2 but we can add that in later), but what of our B direction? Well, it's given by figuring out $\vec B_0 = \frac{\vec k}{\omega} \times \vec E_0$. k is z-directional in our situation, and as we've already solved the magnitude, we get a direction of: $\hat B = \hat z \times \hat y = - \hat x$

So we now have an amplitude of $\frac{k}{\omega}E_0$ and a direction of $-\hat x$. What of the form of the function itself?

Well, we must satisfy $\vec \nabla \times \vec E = -\frac{\partial \vec B}{\partial t}$

The cross product of E is...

$$ -\frac{\partial}{\partial z} 2 E_0 sin(kz)cos(\omega t) \hat x  $$

Which is already in the right direction. Taking the z-derivative gives us...

$$ - 2 k E_0 cos(kz)cos(\omega t) \hat x $$

Taking the indefinite time integral...

$$ - 2 \frac{k}{\omega} E_0 cos(kz)sin(\omega t) \hat x = - \vec B $$
$$ \Rightarrow 2 \frac{k}{\omega} E_0 cos(kz)sin(\omega t) \hat x = \vec B $$

So our equations are: 

$$ 2 E_0 sin(kz)cos(\omega t) \hat y = \vec E$$

$$ 2 \frac{k}{\omega} E_0 cos(kz)sin(\omega t) \hat x = \vec B $$

But wait, didn't we say that B was in the minus x direction? Why is it positive here? 

This would be because we are actually dealing with a superposition of two waves. Each individual wave would behave slightly differently. For instance, if we converted each individual E-wave into a B-wave seprately by taking the curl of E and then taking the time derivative...

$$ B_0 (sin(kz - \omega t) - sin(kz + \omega t)) \hat x$$
$$ = B_0 (sin(kz)cos(\omega t) - cos(kz)sin(\omega t) - sin(kz)cos(\omega t) - cos(kz)sin(\omega t)) \hat x$$
$$ = 2 B_0 cos(kz)sin(\omega t) \hat x $$

Note how the different signs induced by the initial taking of the time derivatives results in the overall change in sign. (And the same answer we started with.) 

Electric energy is given by $ \frac{\epsilon_0 \vec E^2}{2} $ and magnetic $ \frac{ \vec B^2}{\mu_0 2} $

Notably, when we time average the squared quantities, only the time-dependent portion of the trig function will go away (1/2 in both cases as it will be a square trig function). The results are...

$$ \langle U_E \rangle = \frac{1}{2} \epsilon_0 2^2 E_0 sin^2(kz) \frac{1}{2} = \epsilon_0 E_0 sin^2(kz) $$

$$ \langle U_B \rangle = \frac{1}{2} \frac{1}{\mu_0} 2^2 \frac{k^2}{\omega^2} E_0 cos^2(kz) \frac{1}{2} = \frac{1}{\mu_0} \frac{k^2}{\omega^2} E_0 cos^2(kz) = \epsilon_0 E_0 cos^2(kz) $$

Note that they are the same besides the trig factors. This is expected, single EM plane waves have equal magnitude in energy density over time. However, since this is a standing wave and they are out of phase, these energies are locked to locations. There exist nodes where the E wave doesn't move at all while the B wave is going up and down regularly. 

Now we need to find the time averaged energy current density. Which is just another way of saying the Poynting Vector.

$$ \vec S = \frac{1}{\mu_0} (\vec E \times \vec B) $$

This cross product is not difficult to evaluate as y cross x is -z.

$$ = - \frac{1}{\mu_0} 4 E_0^2 \frac{k}{\omega} sin(kz)cos(kz)sin(\omega t)cos(\omega t) \hat z $$

This is the instantaneous Poynting Vector. It'll oscillate back and forth form positive z to negative z. Naturally this means that over time average we have...

$$ \vec S = 0 $$

Which makes sense, the wave isn't moving overall. 

So we struggled a lot with this one, trying very very hard to make the time dependence vanish since we needed to test at various z-levels. This turned out not to be the answer. We found this out because we physically started plotting out the waves. Here is the electric field with respect to time at z=0.



Looks circularly polarized. However, if we adjust z...



We get a different *elliptical* polarization. So the polarizations are going to be inherently complex. 

Which means what we need to do is not try to *remove* the time dependence, but *pull it out* of the relation. This can be done, if we do a step that seems rather out of left field: pull out a negative sign from the y-polarized wave.

$$ E_0 sin(kz - \omega t) = -E_0 sin(-kz + \omega t) $$

Now wecan write out everything incomplex form.

$$ E_0 ( i e^{i(-kz+\omega t)} \hat y + e^{i(kz+\omega t)} \hat x) = E_0 e^{i\omega t} (i e^{-ikz} \hat y - e^{ikz} \hat x) $$

The polarization is the direction. Which... isn't really easy to see from $i e^{-ikz} \hat y - e^{ikz} \hat x$, so we tried to split into trig components, but that just made it worse. Perhaps there is a reason we are told to find it for specific values of z and not all z. Very well, we shall start plugging numbers in.

z=0 produces...

$$ E_0 e^{i\omega t} (i \hat y - \hat x) $$

Which is a type of circular polarization. 

Now, we need $\lambda$ in terms of k. $\lambda$ is a wavelength, so a full $\lambda$ should set the value of kz to $2\pi$, so thus $\lambda = 2\pi/k$. 

The next one we care about is $\lambda / 8 = \pi/4k$. So the polarization is $i e^{-i\pi/4} \hat y - e^{i\pi/4} \hat x$ or

$$ i (\frac{\sqrt{2}}{2} - i\frac{\sqrt{2}}{2}) \hat y - (\frac{\sqrt{2}}{2} + i\frac{\sqrt{2}}{2}) \hat x $$
$$ = (i \frac{\sqrt{2}}{2} + \frac{\sqrt{2}}{2}) \hat y - (\frac{\sqrt{2}}{2} + i\frac{\sqrt{2}}{2}) \hat x $$

These values are actually correlated to each other, so this is actually a specific polarization! When y is positive x is negative, and vice versa, so we have a $-\hat x + \hat y$ polarization direction. 

Next we have $\lambda / 4$. This evaluates as purely imaginary, giving us

$$ (\hat y - i\hat x) $$

Which is just the opposite circular wave to the first case.

$3 \lambda / 8$ is very similar to $\lambda / 8$ one, but the sign on the constant term is flipped.

$$ (- i \frac{\sqrt{2}}{2} + \frac{\sqrt{2}}{2}) \hat y - (-\frac{\sqrt{2}}{2} + i\frac{\sqrt{2}}{2}) \hat x $$

Which correlates to $\hat x + \hat y$, a specific polarization.

Now, what of $\lambda/2$? Well... it's the same as 0, except the signs are flipped. But a sign flip in a polarization is worthless, +z and -z are still the same plane for the wave to oscillate in. 

$$ (i\hat y - \hat x) : (-i\hat y + \hat x) $$

Which is a circular polarization going the same (curl) direction as z=0. Strictly speaking the sign flip dictates where we "start" at t=0. 




Comparing the actual waves at z=0 and z=$\lambda/2$. The polarization is the same, but the start position is different.

$$ E =2 E_0 [ - sin(kxcos\theta)sin(kysin\theta)cos(\omega t)  + cos(kxcos\theta)sin(kysin\theta)sin(\omega t) ] \hat z $$

Time averaging removes a 2.

$$ = 2 E_0^2 [ sin^2(kysin\theta) + sin^2(kysin\theta) ] $$
$$ = 4 E_0^2 sin^2(kysin\theta)$$

And so the time average electric energy is

$$ \epsilon_0 2 E_0^2 sin^2(kysin\theta)$$

Now we turn to the magnetic field. 

$$ \vec B = 2 \frac{k}{\omega} E_0 [ - cos(kxcos\theta)sin(kysin\theta)sin(\omega t) + sin(kxcos\theta)sin(kysin\theta)cos(\omega t) ] \hat y \\
