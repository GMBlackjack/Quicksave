# Problem 1 \[Back to [top](#toc)\]
$$\label{P1}$$

*Consider an infinite slab of material of thickness t with a hole of radius R drilled through it. Say the material has a magnetization $\vec M = M_0 \frac{R}{\rho} \hat \rho$ where $\rho$ is the distance from the hole's axis.*

*a) What is the magnetic field everywhere along the axis of the hole?*

So, one might think this is a hard problem, but what we actually need to solve for is something simpler than a plane with a hole. We simply need to find the magnetic field of a magnetized infinite plane, and then subtract the magnetic field of a cylinder! In fact, we only need it along an axis, which is actually quite nice. 

First of all, the magnetic field of the magnetized cylinder. We can't just use our previous work directly since the magnetization is not actually constant, but varies with radius. First, we need to check for bound magnetic charges.

$$ - \nabla \cdot \vec M = 0 $$

Because the magnetization is $1/\rho$ which when multiplied by $\rho$ becomes something with a zero derivative. However, we also have the fake surface charges to worry about, and those are given by

$$ \sigma = \vec M \cdot \hat n = \pm  M_0 \frac{R}{\rho} \hat z$$

These calculations would also apply for the infinite slab as well as the cylinder. 

Now, the problem is this charge density isn't constant, so we can't just use the "electric field along the axis of a charged disc" result, we have to makea  new one for a charged disc that varies with radius. However, as it varies *only* with radius, the electric field for a *ring* can still be used.

$$ \vec E = \frac{Q}{4\pi \epsilon_0} \frac{z}{(z^2+R^2)^{3/2}} \hat z $$

So for the H-field...

$$ \vec H = \frac{Q}{4\pi} \frac{z}{(z^2+R^2)^{3/2}} \hat z $$

Keep in mind we are currently measuring from the center of the ring, not from the actual center of the system, would exist t/2 below this point. What we need is an integral we can perform over not z, but R--since the radius is what we will be varying. One thing we need to be careful about is the total charge: Q for the ring is $\lambda 2 \pi R$, but we have $\sigma$ instead of $\lambda$. Which means that if we were to consider the surface charge to be a bunch of added togehter line charges, then $\lambda = \pm M_0 R/\rho \hat z$ *at a specific value of $\rho$.* Since at every specific value of rho the density is constant, however, we don't have to worry about re-integrating over the angle, the total charge density over any ring would be "Q" = $M_0 \frac{R}{\rho} 2 \pi \rho \hat z$ which means, that as far as we're concerned, the H-field for an individual ring is:

$$ \vec H = \frac{M_0 \frac{R}{\rho} 2 \pi \rho}{4\pi} \frac{z}{(z^2+\rho^2)^{3/2}} \hat z $$
$$ = \frac{M_0 R}{2} \frac{z}{(z^2+\rho^2)^{3/2}} \hat z $$

Now to add all of them together we integrate over $\rho$. IF we were doing a disc, it would be from 0 to R, and for the plane it woudl be 0 to infinity. 

$$ \frac{z M_0 R}{2} \int \frac{1}{(z^2+\rho^2)^{3/2}} \hat z d\rho $$

Now, surely this integral has been done before somewhere? 

It was, on page 5 of handout 20. From it we were able to reconstruct the following relationship:

$$ \frac{d}{d\rho} \frac{\rho}{\sqrt{z^2+\rho^2}} = \frac{1}{\sqrt{z^2+\rho^2}} - \frac{\rho^2}{(z^2 + \rho^2)^{3/2}} = \frac{z^2}{(z^2+\rho^2)^{3/2}}$$

Which means our integral would have an extra factor of $z^2$ in it. 

$$ \frac{z M_0 R}{2} \left[ \frac{1}{z^2} \frac{\rho}{\sqrt{z^2+\rho^2}} \right]_0^? \hat z $$

Given the $\rho$ in the numerator we know the lower bound, 0, evaluates to nothing. Which is to be expected since a ring with radius zero wouldn't provide anything. 

For the disc, this evaluates to:

$$ \vec H_d = \frac{M_0 R}{2z} \frac{R}{\sqrt{z^2+R^2}} \hat z $$

And for the infinite plane...

$$ \vec H_p = \frac{M_0 R}{2z} \hat z $$

Since for large $\rho$ the fraciton goes to 1. Naturally the plane has more field than the disc, which is to be expected. Both fields lower with distance away from the object itself, which is also good. The field would be negative for the negatively charged versions.

Now that we have the formulas, we need to find the H-field for the cylinder and the slab, which will involve minor adjustments to the variable z to account for locaiton. Chooseing the z=0 point to be the center of the slab, we end up with...

$$ \vec H_c = \frac{M_0 R}{2(z-t/2)} \frac{R}{\sqrt{(z-t/2)^2+R^2}} - \frac{M_0 R}{2(z+t/2)} \frac{R}{\sqrt{(z+t/2)^2+R^2}} \hat z$$

Note the signs. The first term is positively charged, but eblow z=t/2 we are below it and thus positive charge woudl be pushed *down*, hence the negative sign. However, if we were above t/2, we would be pushed *up.* This is why we haven't used absolute value signs, since the direction changes based on position, and naturally there is a divergence *at* the charge itself. The negatively charged side, meanwhile, will pull down unless z is less than -t/2, in which case it will start to pull up--that is, assuming our "test charge" is positive, which doesn't even really make much sense since our charge is fake anyway, but it's a way to keep track of the signs.

There is similar justification for a set of two plates that make up a slab.

$$ \vec H_s = \frac{M_0 R}{2(z-t/2)} - \frac{M_0 R}{2(z+t/2)} \hat z $$


Now we can find the magnetic field of the whole slab and the cylinder just by $\vec B = \mu_0 \vec M + \mu_0 \vec H$

$$ \vec B_c = M_0\mu_0\frac{R}{\rho}  + \frac{M_0 R\mu_0}{2(z-t/2)} \frac{R}{\sqrt{(z-t/2)^2+R^2}} - \frac{M_0 R\mu_0}{2(z+t/2)} \frac{R}{\sqrt{(z+t/2)^2+R^2}} \hat z $$

$$ \vec B_s = M_0\mu_0 \frac{R}{\rho} + \frac{M_0 R \mu_0}{2(z-t/2)} - \frac{M_0 R \mu_0}{2(z+t/2)} \hat z $$

Now, one might complain that our notation is a bit off: after all, the magnetization doesn't exist outside the material. Well, if we just let the magnetization be a function dependent on z that only exists on [-t/2, t/2], then our form still works. The much BIGGER problem is the fact that the magnetization is *undefined* on the axis, which is kind of where we need it to be defined! However, we are fortunate that the magnetization in the center does not *actually exist*. Once we subtract the cylindrical magnetic field from the slab's magnetic field, our problem with the undefined magnetization goes away because the magnetization *entirely* cancels. 

$$ \vec B_{final} = \frac{M_0 R \mu_0}{2(z-t/2)} - \frac{M_0 R \mu_0}{2(z+t/2)} - \frac{M_0 R\mu_0}{2(z-t/2)} \frac{R}{\sqrt{(z-t/2)^2+R^2}} + \frac{M_0 R\mu_0}{2(z+t/2)} \frac{R}{\sqrt{(z+t/2)^2+R^2}} \hat z $$

The fact that the magnetization cancels out should not be a surprsie since there is not acutally any magnetized material on the axis! So it does not matter that such a magnetization would cause a discontinuity, everything neatly removes itself and we have an answer. The result above is a suitable answer, but we can simplify it considerably.

$$ \vec B_{final} = \frac{M_0 R \mu_0}{2} \left[ \frac{1}{(z-t/2)} - \frac{1}{(z+t/2)} - \frac{1}{(z-t/2)} \frac{R}{\sqrt{(z-t/2)^2+R^2}} + \frac{1}{(z+t/2)} \frac{R}{\sqrt{(z+t/2)^2+R^2}} \right] \hat z $$

At this point we could try to simplify even further, but we would have to choose between grouping together the 1/z terms or the radical terms, and there appears to be no distinct advantage of one over the other, so we shall leave it like this. We can say that the form is sensible--the larger contributions of the plane are reduced by the smaller contributions of the disc-shaped holes, and the upper plane is positively "charged" and the bottom is "negative," and with no magnetized material the B-field is equal to the H-field save for the factor of $\mu_0$

*b) Say the small magnetic dipole is constrained to only move along the axis of the hole and is forced to keep its dipole moment aligned upwards along the axis of the hole (i.e. $\vec m = m\hat z$). What is the net magnetic force on the dipole if it is located at the top, middle, and bottom of the hole? Could any of these positions be a stable equilibrium location for the dipole?*

It looks like we have discontinuities at z=t/2, but actually we conveniently don't, though it's not obvious that that's the case. To show why, let's think about what happens to the radicals when the $z\pm t/2$ term goes to zero--we get R/R, which is 1. This means every radical term is left with only its leading $\frac{1}{z\pm t/2}$ term, which will *always* cancel with the term that originally came from the plane. Sure, both terms do diverge, but they diverge *equally* and thus cancel at the boundary itself. Therefore...

$$ \vec B_{final}(t/2) = \frac{M_0 R \mu_0}{2} \left[ - \frac{1}{t} + \frac{1}{t} \frac{R}{\sqrt{t^2+R^2}} \right] \hat z $$
$$ \vec B_{final} = \frac{M_0 R \mu_0}{2} \left[ -\frac{4}{t} + \frac{4}{t} \frac{R}{\sqrt{t^2/4+R^2}}  \right] \hat z $$
$$ \vec B_{final}(-t/2) = \frac{M_0 R \mu_0}{2} \left[ -\frac{1}{t} + \frac{1}{t} \frac{R}{\sqrt{t^2+R^2}} \right] \hat z $$

Note that no matter what we do, the field always points down on the range. This is exactly what we would expect, for being below the "positive" end would have downward pointing field arrows, as would anything above the "negative" end, so in the realm where we are between them the field lines better point downward, even if we are not actually "between" anything real and are in fact in a hole. 

Now we are prepared to find the force on a dipole due to a magnetic field. 

For a point dipole, the force is:

$$ \vec F = \nabla (\vec m \cdot \vec B) $$

Now, since both of the vectors are in the z direction, this becomes:

$$ \vec F = \frac{d}{dz} mB \hat z $$

Now, we note that our specific values of the magnetic field won't help us here, since we need *change* in the z direction to get anything done! However, since the magnetic moment m istelf is constant, all we need to do is take the z-derivative of the magentic field. As it is a derivative and not an integral we can just do this manually without scouring the notes for a place it was done before.

$$ \vec F = m\frac{M_0 R \mu_0}{2} \left[ -\frac{1}{(z-t/2)^2} + \frac{1}{(z+t/2)^2} + \frac{1}{(z-t/2)^2} \frac{R}{\sqrt{(z-t/2)^2+R^2}} \\ - \frac{1}{(z+t/2)^2} \frac{R}{\sqrt{(z+t/2)^2+R^2}} + \frac{R}{((z-t/2)^2+R^2)^{3/2}} -  \frac{R}{((z+t/2)^2+R^2)^{3/2}} \right]   \hat z $$

The first four terms behave very similarly to B itself at the three tested points, while the last two terms are additions that come from the product rule for derivatives. What we see above is the force at every point along the axis. If we evaluate it at our three poitns of interest keeping in mind what we learned from the cancelations in B, we get...

$$ \vec F(t/2) = m\frac{M_0 R \mu_0}{2} \left[ \frac{1}{t^2} - \frac{1}{t^2} \frac{R}{\sqrt{t^2+R^2}} - \frac{R}{(t^2+R^2)^{3/2}} + 1 \right]   \hat z $$

$$ \vec F(0) = 0 $$

$$ \vec F(-t/2) = m\frac{M_0 R \mu_0}{2} \left[ - \frac{1}{t^2} + \frac{1}{t^2} \frac{R}{\sqrt{t^2+R^2}} + \frac{R}{(t^2+R^2)^{3/2}} - 1 \right]   \hat z $$

Note that while $\vec B$ is not zero at z=0, the derivative *is* entirely due to the fact that the suddenly squared terms all cancel out, rather than being additive due to sign changes. Note also that the direction of the force changes even though the direction of the B-field line does not! This is due to the rate of change adjusting, so the point in the center is either a maxima or a minima for the magnetic field. From the calculation of the magnetic field itself we can see pretty clearly that it is a maxima at this point, that is, the field is strongest in the center. (Which is to be expected, naturally). 

So yes, the center of the hole is an equilibrium point! This makes sense, as if we consider the "forces" as a result of charge, there is an equal amount of charge above and below, so it would cancel out. 

Another question someone might ask is if it is a *stable* equilibrium? Well, we can examine that by looking at the signs of the forces at the ends, t/2 and -t/2. Both are equal in magnitude but opposite in direction. We note that the sign at these points is actually guaranteed--$1/t^2$ always dominates over the second term, and 1 always dominates over the third term, no matter what, thus (assuming smooth continuity of force, which is reasonable) the upper force is always positive and the lower force is always negative, meaning that our dipole really wants to be flung upward or downward at high speed, so this is an *unstable* equilibrium. 

However, this is not strictly part of the problem we were asked to solve, it was just a curiosity. In summary, the magnetic field in the axis is what we found in part a), the forces at the ends t/2 and -t/2 are equal but opposite, and the center has a net zero force and serves as an equilibrium point. As a bonus, the equilibrium is an unstable equilibrium, this isn't one of those fancy levitating magnets. 

Notably, if we flipped the direction of the dipole, the forces would reverse, and then we *would* have stable equilibrium. Of course, we are not permitting the dipole to rotate, so a realistic situation could mess things up with potential torques.

*Consider a rectangular toroidal solenoid with radii $R_a$ and $R_b$ and height h with N loops carrying a current I. A material with a suspectibility $\chi$ fills the region between radii $R_1$ and $R_2$ within this toroid. What is the magnetic field and bound current distribution everywhere in this system?*

Here's a quick image we made to illustrate what this looks like from a cross-section.



Note: far later in the problem we realize our current direction model up there puts the magnetic field in the $-\hat\phi$ direction, which while not incorret, does cause some sign confusion. The actual answer will assume default direction of $\hat\phi$. 

Let us begin by first finding the homework where we solved for a rectangular toroid already. This would be problem 3 in homework 10, which yes we know we did incorrectly but we still have the answers for it. (Which are curiously labeled homework 11, interesting). 

From this, we have the magnetic field inside the solenolid without the material in it. 

$ B = \frac{\mu_0 N I}{2 \pi \rho} $, where $\rho$ is radius measured from the axis. Notably this field only exists inside the solenoid, outside it is neglegible. In fact, we can derive this rather simply by Ampere's law--a circle in the toroid's inteiror has precisely NI current loops going through it, but if we were to extend the circle to encompass the entire toroid, there would also be -NI current, canceling it all to zero, thus the magnetic field only exists in the toroid interior. 

Continuing to pretend like our magnetizable material isn't there, we find that there is only free current, no magnetization, and that the H-field is $H = \frac{N I}{2 \pi \rho}$

The direction of these fields is somewhat in question, as they are certaionly $\phi$ restricted, but the sign depends on the direction of I and which way we consider the current to be flowing. Since it is left ambiguous, we shall assume the simple case where the $\phi$ direction is positive. 

$$ \vec B_t = \frac{\mu_0 N I}{2 \pi \rho} \hat \phi$$
$$ \vec H_t = \frac{N I}{2 \pi \rho} \hat \phi $$

From this, we can now find the magnetization of our situation, $\vec M_\chi = \frac{\chi}{\mu} \vec B$. (Note that the ratio is reversed intentionally, the fraciton in the notes was backward as discussed in the email.) 

Just to be eplicit, we can always state $\mu$ in terms of $\chi$

$$ \mu = \mu_0 (1+\chi) $$

So we are not introducing any unknowns here. 

This means that our induced magnetization in our toroidal material is

$$ \vec M_\chi = \frac{\chi}{\mu} \frac{\mu_0 N I}{2 \pi \rho} \hat \phi $$

Immediately we can see that there will be no fake surface charge on the toroid, the normal vectors point in the z and radial directions, never $\phi$, so the surfaces are clean. The VOLUME, however... 

$$ \rho = -\nabla \cdot \frac{\chi}{\mu} \frac{\mu_0 N I}{2 \pi \rho} \hat \phi = 0$$ 

Because even though we have $\phi$ directionality, we have no $\phi$ dependence. Thus, there is no magnetic charge, and thus the toroid provides no H-field. Meaning outside the magnetic material, the magnetic field is unchanged. 

This is not the case *inside* the material, however. That said since there is no H-field we just have:

$$ \vec B_\chi = \frac{\chi}{\mu} \frac{\mu^2_0 N I}{2 \pi \rho} \hat \phi $$

Now this can probably be stated in some form similar to $\frac{\chi}{\chi+2}$, but given the reversal of the fraction we don't exactly trust our work on **Problem 1b** in HW 11 so we'll not state it in those terms and just have the annoying $\mu_0^2$ term sitting there. It is just a constant, after all.

Note that there is still no bound CHARGE, just that the magnetic field itself will be changing inside the material. We actually have our answer for the magnetic field everywhere, now. 

$$ \vec B_{out} = 0 $$
$$ \vec B_{intoroid} = \frac{\mu_0 N I}{2 \pi \rho} \hat \phi $$
$$ \vec B_{inmaterial} = \frac{\mu_0 N I}{2 \pi \rho} + \frac{\chi}{\mu} \frac{\mu^2_0 N I}{2 \pi \rho} \hat \phi $$

Where "in toroid" is specifically in the area inside the toroid and NOT inside the material. 

To finish the problem, we need to find the bound current. We note that the bound current would be whatever current would produce the magnetic field just inside the material. Which would be like going in reverse: if a current of I produces $\frac{\mu_0 N I}{2 \pi \rho} \hat \phi$, then in order to produce $ \frac{\chi}{\mu} \frac{\mu^2_0 N I}{2 \pi \rho} \hat \phi  $ we need a bound current of...

$$ I_b = \frac{\chi}{\mu} \mu_0 I $$

There is one small problem with this, though. This answer assumes the magnetic material is a magnetic toroid all its own with the sane number of loops as the actual toroid outsdie it. Physically, there obviously isn't a toroid here, it's a solid magnetic ring with a surface current spinning around it. One might think the current is constant, but this actually isn't tha case as the outer edge of the toroid is *bigger* than the inner edge, so the distribution of current must change with $\rho$. So to fully answer the question, we'll need to find the actual current distributions. 

Since the inner toroid is rectangular, we have four surfaces: inner $R_2$, outer $R_1$, and both the top and bottom. If this were a toroid with loops, the surface current could be represented by $\frac{NI_b}{2\pi R_2}$ on the interior, and $\frac{NI_b}{2\pi R_1}$ on the exterior. Notably on a larger radius, the density is lower, *as it should be*. Changing height would not change the current density. However, changing radius on the top and bottom will--but that can easily be represented by $\frac{NI_b}{2\pi \rho}$. 

With this, we technically have the current distributions, but we have not assigned any *signs* to them. Let's take the phi-direction as the way we expect the magnetic field to be going (as opposed to negative phi) and then we get...

$$ K_{r=R_2} = \frac{NI_b}{2\pi R_2} \hat z$$
$$ K_{r=R_1} = - \frac{NI_b}{2\pi R_1} \hat z$$
$$ K_{z=h/2} = \frac{NI_b}{2\pi \rho} \hat \rho$$
$$ K_{z=-h/2} = - \frac{NI_b}{2\pi \rho} \hat \rho$$

$$ I_b = \frac{\chi}{\mu} \mu_0 I $$

Because the bound currents have to go in a full loop. And every one of these, if integrated over any radius, will produce a total current of $NI_b$ which is what we expect. 

Notably we can model the surface current for the actual toroid as well, just replace $I_b$ with I and the 12 subscripts with ab. But as it has actual physical loops saying it has an actual current density is suspect at best. That said, in the interests of being complete in the answer, here are the free currents:

$$ K_{r=R_b} = \frac{NI}{2\pi R_b} \hat z$$
$$ K_{r=R_a} = - \frac{NI}{2\pi R_a} \hat z$$
$$ K_{z=h/2} = \frac{NI}{2\pi \rho} \hat \rho$$
$$ K_{z=-h/2} = - \frac{NI}{2\pi \rho} \hat \rho$$

Keep in mind that the currents all trace rectangles and don't extend beyond the bounds that meet with the other currents. The free currents are on the toroid, the bound on the magnetizable material.

Physically, the material here is providing an amplification for the magnetic field inside of it, which is genreally what is expected. Furthermore, its shape as a ring also isn't surprising: it'll act as another toroid and only produce magnetic field inside itself. This leads to the interesting conclusion that there is no way to tell from otuside if a toroid is filled with a magnetizable material or not, as all the magnetic field boosting occurs inside the material.

This leads to the obvious conclusion that putting a magnetizable material inside a toroid doesn't really have much in the way of practical bennefits, we generally want to boost magnetic fields *outside* the magnetizable materials.
